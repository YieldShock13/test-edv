```python
# ============================================================
# 0) INSTALL + IMPORTS
# ============================================================
!pip install hmmlearn pandas_datareader

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from hmmlearn.hmm import GaussianHMM
from pandas_datareader import data as pdr

# Make printed probabilities readable
pd.options.display.float_format = "{:.3f}".format

# Simple Orca / Bloomberg-ish style for plots
plt.rcParams.update({
    "figure.facecolor": "#0e1117",
    "axes.facecolor":   "#0e1117",
    "axes.edgecolor":   "#3a3a3a",
    "axes.labelcolor":  "white",
    "xtick.color":      "white",
    "ytick.color":      "white",
    "grid.color":       "#3f3f3f",
    "figure.figsize":   (12, 4)
})

def add_watermark():
    ax = plt.gca()
    ax.text(
        0.5, 0.5, "Orca's Strategies",
        transform=ax.transAxes,
        fontsize=32,
        color="white",
        alpha=0.06,
        ha="center", va="center",
        fontweight="bold",
    )

# ============================================================
# 1) FEATURE MATRIX FOR HMM  (ASSUMES macro_state_q EXISTS)
# ============================================================

# Keep rows with all required inputs
features_df = macro_state_q.copy().dropna(subset=[
    "output_gap_pct",
    "inflation_gap",
    "policy_gap",
    "stress_composite",
])

# Standardize ONLY the two gap variables to avoid scale dominance
scaler = StandardScaler()
features_df[["output_gap_z", "inflation_gap_z"]] = scaler.fit_transform(
    features_df[["output_gap_pct", "inflation_gap"]]
)

# Final feature matrix X: 4 variables
X = features_df[[
    "output_gap_z",      # growth gap (z-score)
    "inflation_gap_z",   # inflation gap (z-score)
    "policy_gap",        # policy stance vs neutral
    "stress_composite"   # financial stress
]].values

dates = features_df.index

# ============================================================
# 2) FIT 4-STATE GAUSSIAN HMM
# ============================================================

n_states = 4
hmm = GaussianHMM(
    n_components=n_states,
    covariance_type="full",
    n_iter=500,
    random_state=42,
)
hmm.fit(X)

# ============================================================
# 3) STATE SEQUENCE + PROBABILITIES
# ============================================================

state_sequence = hmm.predict(X)
state_probs = hmm.predict_proba(X)  # shape (T, 4): P(state=k | X_t)

prob_cols = [f"prob_regime_{k}" for k in range(n_states)]
probs_df = pd.DataFrame(state_probs, index=dates, columns=prob_cols)

regime_df = pd.DataFrame({"regime": state_sequence}, index=dates)
regime_df = pd.concat([regime_df, probs_df], axis=1)

# Combine back into macro_regime_q
macro_regime_q = pd.concat([macro_state_q, regime_df], axis=1)

print("=== STATE MACRO PROFILES (z-gaps, policy, stress) ===")
state_means = []
for k in range(n_states):
    mask = (state_sequence == k)
    state_means.append(
        features_df.loc[mask, [
            "output_gap_z",
            "inflation_gap_z",
            "policy_gap",
            "stress_composite"
        ]].mean()
    )
state_means_df = pd.DataFrame(state_means)
state_means_df.index.name = "state"
print(state_means_df)

print("\n=== LAST 5 ROWS OF macro_regime_q (WITH REGIMES) ===")
print(macro_regime_q.tail(5))

# ============================================================
# 4) CURRENT REGIME SNAPSHOT
# ============================================================

current = macro_regime_q.iloc[-1].copy()
current_probs = current[prob_cols]

current_snapshot = pd.DataFrame(
    {
        "value": [
            current.name,                   # date
            int(current["regime"]),         # regime index
            float(current_probs.max()),     # highest regime prob
            current_probs.idxmax(),         # which regime
            current["output_gap_pct"],
            current["inflation_yoy"],
            current["policy_gap"],
            current["stress_composite"],
        ]
    },
    index=[
        "date",
        "regime",
        "max_regime_prob",
        "max_regime_prob_state",
        "output_gap_pct",
        "inflation_yoy",
        "policy_gap",
        "stress_composite",
    ],
)

print("\n=== CURRENT REGIME SNAPSHOT ===")
print(current_snapshot)

# ============================================================
# 5) NBER RECESSION DUMMY (USRECQ FROM FRED)
# ============================================================

start_date = macro_regime_q.index.min()
end_date   = macro_regime_q.index.max()

# USRECQ is a quarterly NBER recession indicator: 1=recession, 0=expansion
usrecq = pdr.DataReader("USRECQ", "fred", start=start_date, end=end_date)
usrecq = usrecq.resample("Q").last()  # ensure quarterly alignment
usrecq = usrecq.rename(columns={"USRECQ": "USREC"})

macro_regime_q = macro_regime_q.join(usrecq, how="left")
macro_regime_q["USREC"] = macro_regime_q["USREC"].fillna(0)

# ============================================================
# 6) VALIDATION: HOW EACH REGIME LINES UP WITH NBER RECESSIONS
# ============================================================

validation_rows = []
for k in range(n_states):
    mask = (macro_regime_q["regime"] == k)
    n_k = mask.sum()
    if n_k == 0:
        pct_rec = np.nan
        pct_exp = np.nan
    else:
        pct_rec = macro_regime_q.loc[mask, "USREC"].mean()
        pct_exp = 1.0 - pct_rec
    validation_rows.append([k, n_k, pct_rec, pct_exp])

validation_df = pd.DataFrame(
    validation_rows,
    columns=["regime", "n_obs", "pct_recession", "pct_expansion"]
)

print("\n=== REGIME vs NBER RECESSION VALIDATION ===")
print(validation_df)

# Interpretation:
# pct_recession close to 1 -> that regime mostly falls in recessions
# pct_recession close to 0 -> that regime mostly falls in expansions

# ============================================================
# 7) REGIME TIMELINE PLOT (WITH ORCA WATERMARK)
# ============================================================

colors = {0: "green", 1: "orange", 2: "red", 3: "blue"}

plt.figure(figsize=(12, 3))
plt.scatter(
    dates,
    state_sequence,
    c=[colors[int(s)] for s in state_sequence],
    marker="s",
    s=45,
    edgecolor="black"
)
plt.yticks(range(n_states), [f"Regime {i}" for i in range(n_states)])
plt.title("Macro Regime Over Time", fontsize=14, fontweight="bold")
plt.xlabel("Date")
plt.grid(True, alpha=0.3)
add_watermark()
plt.tight_layout()
plt.show()









```
