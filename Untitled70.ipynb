```python
# ================================
# DE 2Y  US 2Y (Monthly, nearest to 7th of each month) + OLS
# Capped to end on 2025-09 (7 Sep 2025 month)
# ================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import statsmodels.api as sm
import os, glob

WATERMARK = "Orca's Capital Strategies"
EXACT_GER = "Germany 2-Year Bond Yield Historical Data (2).csv"
EXACT_US  = "United States 2-Year Bond Yield Historical Data (1).csv"
END_MONTH = pd.Period("2025-09", freq="M")  # ensure we end with Sep-2025

# ---------- file resolution ----------
def resolve_path(filename):
    for p in [f"/content/{filename}", f"/mnt/data/{filename}", filename]:
        if os.path.isfile(p):
            return p
    # last resort: search
    for base in ["/content", "/mnt/data", "."]:
        hits = glob.glob(os.path.join(base, "**", filename), recursive=True)
        if hits:
            return hits[0]
    raise FileNotFoundError(f"File not found: {filename}")

ger_path = resolve_path(EXACT_GER)
us_path  = resolve_path(EXACT_US)

# ---------- parsing helpers ----------
def parse_price(series: pd.Series) -> pd.Series:
    s = series.astype(str).str.strip().str.replace("%","",regex=False)
    # detect decimal comma
    sample = s.sample(min(100, len(s)), random_state=0)
    comma_no_dot = ((sample.str.contains(",")) & ~sample.str.contains(r"\.")).mean()
    if comma_no_dot > 0.6:
        s = s.str.replace(".", "", regex=False)   # thousands dot
        s = s.str.replace(",", ".", regex=False)  # decimal comma -> dot
    else:
        s = s.str.replace(",", "", regex=False)   # thousands comma
    return pd.to_numeric(s, errors="coerce")

def load_series_raw(path: str, name: str) -> pd.Series:
    df = pd.read_csv(path)
    df.columns = [c.strip() for c in df.columns]
    if "Date" not in df.columns or "Price" not in df.columns:
        raise ValueError(f"'Date' and 'Price' required in {path}. Found: {df.columns.tolist()}")

    # Robust date parse (try both day-first and month-first; choose later max)
    s = df["Date"].astype(str)
    d1 = pd.to_datetime(s, dayfirst=True,  errors="coerce")
    d2 = pd.to_datetime(s, dayfirst=False, errors="coerce")
    dd = d1 if d1.max() >= d2.max() else d2

    df["Date"] = dd
    df["Price"] = parse_price(df["Price"])
    df = df.dropna(subset=["Date","Price"]).sort_values("Date").set_index("Date")
    # Keep original dates (monthly files may already be 1 per month, but we allow multi)
    return df["Price"].rename(name)

# ---------- anchor to the 7th (nearest in-month) ----------
def select_nearest_7th_in_month(series: pd.Series, target_day: int = 7) -> pd.DataFrame:
    """
    For each calendar month present in 'series', pick the row whose DATE (within that month)
    is nearest to the target_day (default 7). Ties resolved by earlier date.
    Returns a DataFrame with:
      - 'value': the selected value
      - 'obs_date': the actual date chosen in that month
      - index: Period('M') month key
    """
    if series.empty:
        return pd.DataFrame(columns=["value","obs_date"])

    df = series.to_frame("value").copy()
    df["month"] = df.index.to_period("M")
    df["day"] = df.index.day

    # For each month, compute absolute distance to target_day and pick min
    def _pick(group):
        group = group.copy()
        group["dist"] = (group["day"] - target_day).abs()
        # sort by distance, then by date asc to break ties by earlier date
        group = group.sort_values(["dist", group.index.name])
        chosen = group.iloc[0:1]  # one row
        out = pd.DataFrame({
            "value": [chosen["value"].iloc[0]],
            "obs_date": [chosen.index[0]]
        }, index=[chosen["month"].iloc[0]])
        return out

    anchored = df.groupby("month", group_keys=False).apply(_pick)
    anchored.index.name = "month"
    return anchored

# ---------- Load & anchor ----------
DE_raw = load_series_raw(ger_path, "DE2Y")
US_raw = load_series_raw(us_path,  "US2Y")

DE_anch = select_nearest_7th_in_month(DE_raw, target_day=7).rename(columns={"value":"DE2Y"})
US_anch = select_nearest_7th_in_month(US_raw, target_day=7).rename(columns={"value":"US2Y"})

# Align by month (inner join), and cap to END_MONTH
panel = DE_anch.join(US_anch, how="inner", lsuffix="_de", rsuffix="_us")
panel = panel.loc[panel.index <= END_MONTH].copy()

# If either series lacks Sep-2025 (END_MONTH) entry, the sample will end earlier by design.
if panel.empty:
    raise RuntimeError("No overlapping months after anchoring to the 7th. Check the files contain overlapping months up to 2025-09.")

# Compute spread & z
panel["Spread_bps"] = (panel["DE2Y"] - panel["US2Y"]) * 100.0
panel["Spread_Z"] = (panel["Spread_bps"] - panel["Spread_bps"].mean()) / panel["Spread_bps"].std(ddof=0)

# For plotting, set a datetime index at month start + 6 days (the 7th nominal anchor)
# but keep the actual chosen obs dates for reference in console.
plot_index = panel.index.to_timestamp(how="start") + pd.Timedelta(days=6)
panel_plot = panel.copy()
panel_plot.index = plot_index

# ---------- OLS: DE2Y ~ US2Y on anchored monthly sample ----------
X = sm.add_constant(panel["US2Y"])
y = panel["DE2Y"]
ols = sm.OLS(y, X).fit()

# ---------- Plots ----------
date_fmt_full = DateFormatter("%Y-%m-%d")

# Chart 1: Spread (bps)  anchored monthly
fig1, ax1 = plt.subplots(figsize=(11, 5.5))
ax1.plot(panel_plot.index, panel_plot["Spread_bps"])
ax1.set_title("Germany 2Y  US 2Y Yield Spread (Monthly, nearest to 7th)")
ax1.set_ylabel("Spread (bps)")
ax1.set_xlabel("Month (anchored to 7th)")
ax1.grid(True, alpha=0.3)
ax1.xaxis.set_major_formatter(date_fmt_full)

last_month = panel.index.max()
last_plot_date = panel_plot.index.max()
last_spread = panel_plot["Spread_bps"].iloc[-1]
ax1.annotate(f"{last_spread:.1f} bps", xy=(last_plot_date, last_spread),
             xytext=(10, 0), textcoords="offset points", va="center")
fig1.text(0.01, 0.01, WATERMARK, fontsize=12, alpha=0.35, ha="left", va="bottom")
plt.tight_layout(); plt.show()

# Chart 2: Z-score  anchored monthly
fig2, ax2 = plt.subplots(figsize=(11, 5.5))
ax2.plot(panel_plot.index, panel_plot["Spread_Z"])
ax2.axhline(0, linewidth=1)
ax2.set_title("Z-Score: Germany 2Y  US 2Y (Monthly, nearest to 7th)")
ax2.set_ylabel("Z-score")
ax2.set_xlabel("Month (anchored to 7th)")
ax2.grid(True, alpha=0.3)
ax2.xaxis.set_major_formatter(date_fmt_full)
ax2.annotate(f"{panel_plot['Spread_Z'].iloc[-1]:.2f}", xy=(last_plot_date, panel_plot['Spread_Z'].iloc[-1]),
             xytext=(10, 0), textcoords="offset points", va="center")
fig2.text(0.01, 0.01, WATERMARK, fontsize=12, alpha=0.35, ha="left", va="bottom")
plt.tight_layout(); plt.show()

# ---------- Console summary ----------
print("Files used:")
print("  GER:", ger_path)
print("  US :", us_path)
print("\nAnchoring rule: nearest in-month date to the 7th (ties -> earlier date).")
print(f"Sample months: {panel.index.min().strftime('%Y-%m')}  {panel.index.max().strftime('%Y-%m')} (capped at 2025-09)")
print("\nLast month details (anchored):")
print("  Month:", last_month.strftime("%Y-%m"))
print("  DE obs date:", panel.loc[last_month, "obs_date_de"].date())
print("  US obs date:", panel.loc[last_month, "obs_date_us"].date())
print(f"  Spread(DEUS): {last_spread:.1f} bps")
print("\nOLS on anchored monthly sample: DE2Y ~ const + US2Y")
print(ols.summary())










```
