```python
# COLAB — 2-5-10 BUTTERFLY (LONG + SHORT) + Z-SCORE (252D rolling) shown for last 10 years
# Uses your uploaded FRED CSVs: DGS2 (1).csv, DGS5 (4).csv, DGS10 (14).csv
# Style: black charts, consistent font, last datapoint labeled.

import os, glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ----------------------------
# STYLE / FONT (set this to match your screenshot font)
# ----------------------------
FONT_NAME = "DejaVu Sans"   # change if you want, e.g. "Arial"
plt.rcParams["font.family"] = FONT_NAME

# ----------------------------
# FIND FILES
# ----------------------------
def find_file(patterns):
    for base in [".", "/mnt/data"]:
        for pat in patterns:
            hits = sorted(glob.glob(os.path.join(base, pat)))
            if hits:
                return hits[0]
    return None

p2  = find_file(["DGS2*.csv"])
p5  = find_file(["DGS5*.csv"])
p10 = find_file(["DGS10*.csv"])

if p2 is None or p5 is None or p10 is None:
    raise FileNotFoundError("Missing one of the files: DGS2*.csv, DGS5*.csv, DGS10*.csv (check uploads).")

# ----------------------------
# LOAD (FRED-style: DATE, VALUE)
# ----------------------------
def load_fred_csv(path):
    df = pd.read_csv(path)
    df = df.rename(columns={df.columns[0]: "Date", df.columns[1]: "Value"})
    df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
    df["Value"] = pd.to_numeric(df["Value"], errors="coerce")
    df = df.dropna().sort_values("Date").drop_duplicates("Date", keep="last")
    return df.set_index("Date")["Value"]

y2  = load_fred_csv(p2).rename("Y2")
y5  = load_fred_csv(p5).rename("Y5")
y10 = load_fred_csv(p10).rename("Y10")

df = pd.concat([y2, y5, y10], axis=1).dropna()

# ----------------------------
# BUTTERFLY (equal weights)
#   BF = 2Y - 2*5Y + 10Y
# ----------------------------
df["BF"] = df["Y2"] - 2.0*df["Y5"] + df["Y10"]
df["LONG_BF"]  = df["BF"]
df["SHORT_BF"] = -df["BF"]

# ----------------------------
# Z-SCORE: 252-trading-day rolling on BF
# Then display only last 10 years of that Z series
# ----------------------------
ROLL = 252
mu = df["BF"].rolling(ROLL).mean()
sd = df["BF"].rolling(ROLL).std(ddof=0)
df["Z"] = (df["BF"] - mu) / sd

end = df.index.max()
start_10y = end - pd.DateOffset(years=10)
df_10y = df.loc[df.index >= start_10y].copy()

# ----------------------------
# PLOTTING
# ----------------------------
def plot_series(s, title, ylab):
    s = s.dropna()
    if s.empty:
        raise ValueError(f"{title}: series is empty after rolling/filters.")

    fig = plt.figure(figsize=(14, 6), dpi=140)
    ax = plt.gca()

    fig.patch.set_facecolor("black")
    ax.set_facecolor("black")

    ax.plot(s.index, s.values, linewidth=1.9)

    last_date = s.index[-1]
    last_val = float(s.iloc[-1])
    ax.scatter([last_date], [last_val], s=55, zorder=5)

    ax.annotate(
        f"{last_date.strftime('%Y-%m-%d')}  {last_val:.4f}",
        xy=(last_date, last_val),
        xytext=(10, 12),
        textcoords="offset points",
        color="white",
        fontsize=10,
        bbox=dict(boxstyle="round,pad=0.25", fc="black", ec="white", alpha=0.65),
        arrowprops=dict(arrowstyle="->", color="white", alpha=0.6),
    )

    ax.set_title(title, color="white", fontsize=14, pad=12)
    ax.set_ylabel(ylab, color="white")
    ax.tick_params(colors="white")
    for spine in ax.spines.values():
        spine.set_color("white")
        spine.set_alpha(0.35)
    ax.grid(True, alpha=0.18)
    plt.tight_layout()
    plt.show()

def plot_z(s, title):
    s = s.dropna()
    if s.empty:
        raise ValueError(f"{title}: Z-score empty (need at least {ROLL} observations).")

    fig = plt.figure(figsize=(14, 6), dpi=140)
    ax = plt.gca()

    fig.patch.set_facecolor("black")
    ax.set_facecolor("black")

    # shading: high positive = light red, low negative = light green
    ax.axhspan(0, 10, facecolor="#ff6b6b", alpha=0.13, zorder=0)
    ax.axhspan(-10, 0, facecolor="#51cf66", alpha=0.13, zorder=0)

    ax.plot(s.index, s.values, linewidth=1.9)
    ax.axhline(0, linewidth=1.0, alpha=0.55)

    last_date = s.index[-1]
    last_val = float(s.iloc[-1])
    ax.scatter([last_date], [last_val], s=55, zorder=5)

    ax.annotate(
        f"{last_date.strftime('%Y-%m-%d')}  z={last_val:.2f}",
        xy=(last_date, last_val),
        xytext=(10, 12),
        textcoords="offset points",
        color="white",
        fontsize=10,
        bbox=dict(boxstyle="round,pad=0.25", fc="black", ec="white", alpha=0.65),
        arrowprops=dict(arrowstyle="->", color="white", alpha=0.6),
    )

    ax.set_title(title, color="white", fontsize=14, pad=12)
    ax.set_ylabel(f"Rolling Z-score ({ROLL}D)", color="white")
    ax.tick_params(colors="white")
    for spine in ax.spines.values():
        spine.set_color("white")
        spine.set_alpha(0.35)
    ax.grid(True, alpha=0.18)
    plt.tight_layout()
    plt.show()

# ----------------------------
# OUTPUT CHARTS
# ----------------------------
plot_series(df_10y["LONG_BF"],  "2-5-10 Butterfly (LONG) — level (last 10y)",  "Yield (pct points)")
plot_series(df_10y["SHORT_BF"], "2-5-10 Butterfly (SHORT) — level (last 10y)", "Yield (pct points)")
plot_z(df_10y["Z"], "2-5-10 Butterfly — Z-score of dislocation (252D roll, last 10y)")
```
