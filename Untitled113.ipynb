```python
# Full fresh script (single cell):
# - Installs arch
# - Detects date/price columns
# - Computes 1-day 95% VaR (Historical, GARCH-filtered, COVID-stressed) in %
# - Plots price, return distributions, VaR comparison
# - ADD-ON: prints and plots GARCH(1,1) conditional vol and today's vol forecast

!pip install arch

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from arch import arch_model

# =========================
# 0. PARAMETERS
# =========================
alpha = 0.05                 # 95% VaR
csv_file = "SP500 (3).csv"   # your file name in Colab

# =========================
# 1. LOAD DATA & DETECT DATE / PRICE
# =========================
df = pd.read_csv(csv_file)

# Detect date column (anything containing 'date', else first column)
date_candidates = [c for c in df.columns if "date" in c.lower()]
date_col = date_candidates[0] if date_candidates else df.columns[0]

df[date_col] = pd.to_datetime(df[date_col], errors="coerce")
df = df.dropna(subset=[date_col])
df = df.sort_values(date_col).set_index(date_col)

# Detect price column
preferred_price_names = [
    "Adj Close", "Adj_Close", "Close", "Price",
    "Last", "Index", "S&P 500", "SP500", "Value"
]
price_col = None
for c in preferred_price_names:
    if c in df.columns:
        price_col = c
        break

if price_col is None:
    # Fallback: pick first mostly-numeric non-date column
    candidate_numeric = None
    for c in df.columns:
        series = (
            df[c]
            .astype(str)
            .str.replace(",", "", regex=False)
            .str.replace(" ", "", regex=False)
        )
        nums = pd.to_numeric(series, errors="coerce")
        if nums.notna().mean() > 0.7:
            candidate_numeric = c
            break
    if candidate_numeric is None:
        raise ValueError("Could not find a numeric price column in the CSV.")
    price_col = candidate_numeric

price_series = (
    df[price_col]
    .astype(str)
    .str.replace(",", "", regex=False)
    .str.replace(" ", "", regex=False)
)
prices = pd.to_numeric(price_series, errors="coerce").dropna()

print(f"Using date column:  {date_col}")
print(f"Using price column: {price_col}\n")

# =========================
# 2. RETURNS (SIMPLE, SO VAR IS IN %)
# =========================
ret = prices.pct_change().dropna()  # simple returns
ret.name = "r"

# =========================
# 3. (A) PLAIN HISTORICAL VAR  FULL SAMPLE
# =========================
q_hist = ret.quantile(alpha)        # 5th percentile return
var_hist_pct = -q_hist * 100        # convert to %

# =========================
# 4. (B) FILTERED HISTORICAL VAR  GARCH(1,1) USING TODAY'S VOL
# =========================
# For GARCH stability, scale returns to percent
ret_garch = ret * 100.0             # now in percentage points

am = arch_model(ret_garch, p=1, q=1, mean="constant", vol="GARCH", dist="normal")
res = am.fit(disp="off")

sigma_t = res.conditional_volatility  # in % units (since ret_garch is %)
sigma_t.name = "sigma_t"

fcst = res.forecast(horizon=1)
sigma_target = float(np.sqrt(fcst.variance.iloc[-1, 0]))  # % vol forecast for next day

print(f"GARCH(1,1) daily vol forecast (next day): {sigma_target:.3f}%\n")

# align with returns (ret, not ret_garch)
data = pd.concat([ret, sigma_t], axis=1).dropna()
data = data[data["sigma_t"] > 0]

# Vol scaling: r*_t = r_t * (_target / _t)
scale = sigma_target / data["sigma_t"]
data["r_filtered"] = data["r"] * scale

q_filt = data["r_filtered"].quantile(alpha)
var_filt_pct = -q_filt * 100  # %

# =========================
# 5. (C) STRESSED VAR  WORST 60-DAY WINDOW IN 2020
# =========================
covid_ret = ret.loc["2020-01-01":"2020-12-31"].copy()
if len(covid_ret) < 60:
    raise ValueError("Not enough 2020 data to define a 60-day COVID stress window.")

roll_vol = covid_ret.rolling(window=60, min_periods=60).std()
stress_end = roll_vol.idxmax()

# Get exact 60-observation window by index position, not calendar days
end_pos = covid_ret.index.get_loc(stress_end)
start_pos = max(0, end_pos - 59)
stress_returns = covid_ret.iloc[start_pos:end_pos + 1]

q_stress = stress_returns.quantile(alpha)
var_stress_pct = -q_stress * 100  # %

# =========================
# 6. SUMMARY TABLE (ALL IN %)
# =========================
results = pd.DataFrame({
    "VaR_Type": [
        "Historical_Full",
        "Filtered_Historical_TodayVol",
        "Stressed_COVID_2020"
    ],
    "VaR_95_percent": [
        var_hist_pct,
        var_filt_pct,
        var_stress_pct
    ]
})

print("1-day 95% VaR (in % of portfolio):")
print(results.to_string(index=False))

print("\nCOVID stress window used:")
print(f"Start: {stress_returns.index[0].date()}  End: {stress_returns.index[-1].date()}")
print(f"Obs in stress window: {len(stress_returns)}")

# =========================
# 7. PLOTS
# =========================

# (1) Price series
plt.figure(figsize=(12, 4))
prices.plot()
plt.title("S&P 500 Price Series")
plt.ylabel("Price")
plt.xlabel("Date")
plt.tight_layout()
plt.show()

# (2) GARCH conditional volatility time series (ADD-ON)
plt.figure(figsize=(12, 4))
sigma_t.plot()
plt.title("GARCH(1,1) Daily Conditional Volatility (%)")
plt.ylabel("Vol (%)")
plt.xlabel("Date")
# highlight last point
last_date = sigma_t.index[-1]
last_vol = sigma_t.iloc[-1]
plt.scatter([last_date], [last_vol])
plt.text(last_date, last_vol,
         f"{last_vol:.2f}%",
         ha="left", va="bottom")
plt.tight_layout()
plt.show()

# (3) Histogram of returns: raw vs filtered vs COVID window
plt.figure(figsize=(12, 5))
plt.hist(ret, bins=80, alpha=0.5, label="Raw returns (full)")
plt.hist(data["r_filtered"], bins=80, alpha=0.5, label="Filtered (today vol)")
plt.hist(stress_returns, bins=40, alpha=0.6, label="COVID stress window")
plt.legend()
plt.title("Return Distributions: Raw vs Filtered vs COVID Window")
plt.xlabel("Daily return")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# (4) Bar chart of VaR in %
labels = ["Historical", "Filtered", "COVID Stress"]
vars_pct = [var_hist_pct, var_filt_pct, var_stress_pct]

plt.figure(figsize=(8, 5))
bars = plt.bar(labels, vars_pct)
plt.title("1-day 95% VaR Comparison (%)")
plt.ylabel("VaR (%)")
for bar, v in zip(bars, vars_pct):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
             f"{v:.2f}%", ha='center', va='bottom')
plt.tight_layout()
plt.show()




```
