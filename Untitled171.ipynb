```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

plt.style.use("seaborn-v0_8-darkgrid")

# ==========================================================
# 1. LOAD + CLEAN (INVESTING.COM FORMAT SAFE)
# ==========================================================

def load_yield(path, name):
    df = pd.read_csv(path)

    df["Date"] = pd.to_datetime(df["Date"])
    df = df[["Date", "Price"]]

    # Clean numeric strings (commas, %)
    df["Price"] = (
        df["Price"]
        .astype(str)
        .str.replace(",", "", regex=False)
        .str.replace("%", "", regex=False)
    )

    df["Price"] = pd.to_numeric(df["Price"], errors="coerce")

    df = df.rename(columns={"Price": name})
    df = df.sort_values("Date")
    return df.set_index("Date")

y3m  = load_yield("United States 3-Month Bond Yield Historical Data (2).csv", "3M")
y2   = load_yield("United States 2-Year Bond Yield Historical Data (4).csv", "2Y")
y5   = load_yield("United States 5-Year Bond Yield Historical Data.csv", "5Y")
y10  = load_yield("United States 10-Year Bond Yield Historical Data (13).csv", "10Y")
y20  = load_yield("U.S. 20-Year Bond Yield Historical Data (1).csv", "20Y")

yields = y3m.join([y2, y5, y10, y20], how="inner").dropna()

# ==========================================================
# 2. FREQUENCY BRANCHING
# ==========================================================

daily = yields.copy()
weekly = yields.resample("W-FRI").last().dropna()

# ==========================================================
# 3. LEVEL STATE
# ==========================================================

def rolling_z(df, window):
    return (df - df.rolling(window).mean()) / df.rolling(window).std(ddof=0)

z_daily = rolling_z(daily, 252)
z_weekly = rolling_z(weekly, 52)

# ==========================================================
# 4. BUTTERFLIES
# ==========================================================

def make_flies(df):
    fly5  = df["5Y"]  - 0.5*df["2Y"] - 0.5*df["10Y"]
    fly10 = df["10Y"] - 0.5*df["2Y"] - 0.5*df["20Y"]
    return pd.DataFrame({"Fly_5": fly5, "Fly_10": fly10})

flies_daily = make_flies(daily)
flies_weekly = make_flies(weekly)

fly_z_daily = rolling_z(flies_daily, 252)
fly_z_weekly = rolling_z(flies_weekly, 52)

# ==========================================================
# 5. MOTION STATE (y)
# ==========================================================

delta_daily = daily.diff().dropna()
delta_weekly = weekly.diff().dropna()

vel_daily = delta_daily.rolling(20).mean()
vel_weekly = delta_weekly.rolling(8).mean()

# ==========================================================
# 6. ROLLING PCA ON y
# ==========================================================

def rolling_pca(df, window):
    explained = []
    last_eigvec = None
    dates = []

    for i in range(window, len(df)):
        sub = df.iloc[i-window:i]
        pca = PCA(n_components=3)
        pca.fit(sub)

        explained.append(pca.explained_variance_ratio_)
        last_eigvec = pca.components_
        dates.append(df.index[i])

    explained = pd.DataFrame(
        explained,
        columns=["PC1", "PC2", "PC3"],
        index=dates
    )

    return explained, last_eigvec

pca_daily_exp, pca_daily_vec = rolling_pca(delta_daily, 252)
pca_weekly_exp, pca_weekly_vec = rolling_pca(delta_weekly, 52)

# ==========================================================
# 7. FINAL DATA POINT LABEL FUNCTION
# ==========================================================

def plot_with_last_label(df, title):
    plt.figure(figsize=(12,5))
    for col in df.columns:
        plt.plot(df.index, df[col], label=col)
        plt.scatter(df.index[-1], df[col].iloc[-1])
        plt.text(
            df.index[-1],
            df[col].iloc[-1],
            f"{df[col].iloc[-1]:.2f}",
            fontsize=9
        )
    plt.title(title)
    plt.legend()
    plt.show()

# ==========================================================
# 8. PLOTS
# ==========================================================

plot_with_last_label(z_daily, "Daily 1Y Rolling Z-Scores (Levels)")
plot_with_last_label(fly_z_daily, "Daily 1Y Rolling Z-Scores (Butterflies)")
plot_with_last_label(vel_daily, "Daily Rolling Drift (20D Mean y)")
plot_with_last_label(pca_daily_exp, "Daily Rolling PCA Explained Variance")

# ==========================================================
# 9. STATISTICS OUTPUT
# ==========================================================

print("\n================ CURRENT DAILY Z-SCORES (LEVELS) ================")
print(z_daily.iloc[-1].round(3))

print("\n================ CURRENT DAILY Z-SCORES (BUTTERFLIES) ================")
print(fly_z_daily.iloc[-1].round(3))

print("\n================ CURRENT DAILY DRIFT VECTOR (bps/day approx) ================")
print(vel_daily.iloc[-1].round(4))

print("\n================ CURRENT DAILY PCA EXPLAINED VARIANCE ================")
print(pca_daily_exp.iloc[-1].round(3))

print("\n================ CURRENT DAILY PCA LAST EIGENVECTORS ================")
print(pd.DataFrame(
    pca_daily_vec,
    columns=daily.columns,
    index=["PC1","PC2","PC3"]
).round(3))

print("\n================ DESCRIPTIVE STATS (DAILY LEVELS) ================")
print(daily.describe().round(3))

print("\n================ DESCRIPTIVE STATS (DAILY BUTTERFLIES) ================")
print(flies_daily.describe().round(3))

```
