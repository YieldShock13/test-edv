```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# =========================================================
# 1) SETTINGS
# =========================================================

END_DATE = "2026-03-04"
ROLL_WINDOW = 180
VAR_LEVEL = 0.95
RF_ANNUAL = 0.04
TRADING_DAYS = 252

# =========================================================
# 2) FILES
# =========================================================

FILES = {
    "pure_hedge": "0P0001K7XM Historical Data (3).csv",
    "high_growth": "0P0001K7XN Historical Data (1).csv",
    "tga": "TGAJ Historical Data.csv",
    "msci_world": "STXWDMJ Historical Data.csv",
    "stxgov": "STXGOVJ Historical Data (7).csv",
    "msci_em": "STXEMGJ Historical Data (2).csv",
    "new_gold": "GLDJ Historical Data (3).csv",
    "truworths": "TRUJ Historical Data (4).csv",
    "wti": "CRU.csv",
}

BENCHMARK = "msci_world"

# =========================================================
# 3) FINAL PORTFOLIO NOTIONALS
# =========================================================

NOTIONALS = {
    "pure_hedge": 25000.0,
    "high_growth": 25000.0,
    "tga": 8500.0,
    "msci_world": 45233.0,
    "stxgov": 28537.0,
    "msci_em": 26459.0,
    "new_gold": 23589.0,
    "truworths": 15481.0,
}

# =========================================================
# 4) DIVIDEND YIELDS TO ADD AS DAILY CARRY
# =========================================================

DIV_YIELDS = {
    "tga": 0.09,
    "truworths": 0.08,
    "stxgov": 0.09,
}

# =========================================================
# 5) HELPERS
# =========================================================

def load_investing_csv(filepath: str, series_name: str) -> pd.DataFrame:
    df = pd.read_csv(filepath)
    df.columns = [c.strip().replace("\ufeff", "") for c in df.columns]

    if "Date" not in df.columns or "Price" not in df.columns:
        raise ValueError(f"{series_name}: missing Date or Price column")

    df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
    df["Price"] = (
        df["Price"]
        .astype(str)
        .str.replace(",", "", regex=False)
        .str.replace(" ", "", regex=False)
    )
    df["Price"] = pd.to_numeric(df["Price"], errors="coerce")

    df = df[["Date", "Price"]].dropna()
    df = df.sort_values("Date").drop_duplicates(subset="Date", keep="last")
    df = df[df["Date"] <= pd.to_datetime(END_DATE)]
    df = df.rename(columns={"Price": series_name})
    return df

def historical_var(x: pd.Series, conf: float = 0.95) -> float:
    x = pd.Series(x).dropna()
    if len(x) == 0:
        return np.nan
    # report as positive loss number
    return -np.percentile(x, (1 - conf) * 100)

def calc_drawdown(index_series: pd.Series) -> pd.Series:
    running_max = index_series.cummax()
    return index_series / running_max - 1.0

def calc_cagr(index_series: pd.Series, periods_per_year: int = 252) -> float:
    s = index_series.dropna()
    if len(s) < 2:
        return np.nan
    years = len(s) / periods_per_year
    if years <= 0:
        return np.nan
    return (s.iloc[-1] / s.iloc[0]) ** (1 / years) - 1

def simple_beta(y: pd.Series, x: pd.Series) -> float:
    tmp = pd.concat([y, x], axis=1).dropna()
    if len(tmp) < 2:
        return np.nan
    yv = tmp.iloc[:, 0]
    xv = tmp.iloc[:, 1]
    vx = xv.var(ddof=1)
    if vx == 0 or np.isnan(vx):
        return np.nan
    return np.cov(yv, xv, ddof=1)[0, 1] / vx

def conditional_beta(y: pd.Series, x: pd.Series, side: str = "down", min_obs: int = 40) -> float:
    tmp = pd.concat([y, x], axis=1).dropna()
    if side == "down":
        tmp = tmp[tmp.iloc[:, 1] < 0]
    elif side == "up":
        tmp = tmp[tmp.iloc[:, 1] > 0]
    else:
        raise ValueError("side must be 'down' or 'up'")

    if len(tmp) < min_obs:
        return np.nan

    yv = tmp.iloc[:, 0]
    xv = tmp.iloc[:, 1]
    vx = xv.var(ddof=1)
    if vx == 0 or np.isnan(vx):
        return np.nan
    return np.cov(yv, xv, ddof=1)[0, 1] / vx

def rolling_beta(y: pd.Series, x: pd.Series, window: int = 180) -> pd.Series:
    out = []
    idx = []
    tmp = pd.concat([y, x], axis=1).dropna()
    for i in range(window, len(tmp) + 1):
        sub = tmp.iloc[i - window:i]
        out.append(simple_beta(sub.iloc[:, 0], sub.iloc[:, 1]))
        idx.append(sub.index[-1])
    return pd.Series(out, index=idx, name="beta")

def rolling_conditional_beta(y: pd.Series, x: pd.Series, window: int = 180, side: str = "down", min_obs: int = 40) -> pd.Series:
    out = []
    idx = []
    tmp = pd.concat([y, x], axis=1).dropna()
    for i in range(window, len(tmp) + 1):
        sub = tmp.iloc[i - window:i]
        if side == "down":
            sub = sub[sub.iloc[:, 1] < 0]
        else:
            sub = sub[sub.iloc[:, 1] > 0]

        if len(sub) < min_obs:
            out.append(np.nan)
        else:
            xv = sub.iloc[:, 1]
            yv = sub.iloc[:, 0]
            vx = xv.var(ddof=1)
            if vx == 0 or np.isnan(vx):
                out.append(np.nan)
            else:
                out.append(np.cov(yv, xv, ddof=1)[0, 1] / vx)
        idx.append(tmp.index[i - 1])
    name = f"{side}_beta"
    return pd.Series(out, index=idx, name=name)

# =========================================================
# 6) LOAD ALL FILES
# =========================================================

series = {}
print("\n=== FILE CHECK ===")
for asset, file in FILES.items():
    df = load_investing_csv(file, asset)
    series[asset] = df
    print(f"{asset:12s} | rows={len(df):4d} | {df['Date'].min().date()} -> {df['Date'].max().date()}")

# =========================================================
# 7) BUILD FULL PRICE PANEL
# =========================================================

prices = None
for asset, df in series.items():
    if prices is None:
        prices = df.copy()
    else:
        prices = prices.merge(df, on="Date", how="outer")

prices = prices.sort_values("Date").set_index("Date").ffill()

required_cols = list(NOTIONALS.keys()) + ["wti"]
prices = prices[required_cols].dropna()

print("\n=== FINAL PANEL ===")
print("Start:", prices.index.min().date())
print("End  :", prices.index.max().date())
print("Rows :", len(prices))

# =========================================================
# 8) PORTFOLIO WEIGHTS
# =========================================================

weights = pd.Series(NOTIONALS, dtype=float)
weights = weights / weights.sum()

print("\n=== FINAL WEIGHTS ===")
print((weights * 100).round(6).astype(str) + "%")

# =========================================================
# 9) DAILY RETURNS
# =========================================================

returns = prices.pct_change().dropna()

# add dividend carry as daily carry
for asset, ann_yield in DIV_YIELDS.items():
    returns[asset] = returns[asset] + ann_yield / TRADING_DAYS

portfolio_assets = list(NOTIONALS.keys())
asset_returns = returns[portfolio_assets].copy()
port_ret = asset_returns.mul(weights, axis=1).sum(axis=1).rename("portfolio")
bench_ret = returns[BENCHMARK].rename("benchmark")
wti_ret = returns["wti"].rename("wti")

# =========================================================
# 10) DAILY ROLLING CORE TESTS
# =========================================================

roll_vol = port_ret.rolling(ROLL_WINDOW).std() * np.sqrt(TRADING_DAYS)
roll_var = port_ret.rolling(ROLL_WINDOW).apply(lambda x: historical_var(x, conf=VAR_LEVEL), raw=False)

roll_beta = rolling_beta(port_ret, bench_ret, ROLL_WINDOW)
roll_down_beta = rolling_conditional_beta(port_ret, bench_ret, ROLL_WINDOW, side="down", min_obs=40)
roll_up_beta = rolling_conditional_beta(port_ret, bench_ret, ROLL_WINDOW, side="up", min_obs=40)

# =========================================================
# 11) PORTFOLIO INDEX, DRAWDOWN, CAGR, SHARPE, CALMAR
# =========================================================

port_index = (1 + port_ret).cumprod()
drawdown = calc_drawdown(port_index)
max_drawdown = drawdown.min()
cagr = calc_cagr(port_index, TRADING_DAYS)

rf_daily = RF_ANNUAL / TRADING_DAYS
excess_ret = port_ret - rf_daily
sharpe = np.sqrt(TRADING_DAYS) * excess_ret.mean() / port_ret.std(ddof=1)
calmar = cagr / abs(max_drawdown) if pd.notna(max_drawdown) and max_drawdown != 0 else np.nan

# =========================================================
# 12) ASSET-LEVEL DIAGNOSTICS
# =========================================================

asset_stats = []
for col in asset_returns.columns:
    asset_stats.append({
        "asset": col,
        "weight": weights[col],
        "daily_ann_vol": asset_returns[col].std(ddof=1) * np.sqrt(TRADING_DAYS),
        "beta": simple_beta(asset_returns[col], bench_ret),
        "down_beta": conditional_beta(asset_returns[col], bench_ret, side="down", min_obs=40),
        "up_beta": conditional_beta(asset_returns[col], bench_ret, side="up", min_obs=40),
    })

asset_stats_df = pd.DataFrame(asset_stats).set_index("asset")

# =========================================================
# 13) RISK CONTRIBUTION
# =========================================================

cov = asset_returns.cov() * TRADING_DAYS
w = weights.values.reshape(-1, 1)
port_var = (w.T @ cov.values @ w).item()
marginal_contrib = cov.values @ w
risk_contrib = (w * marginal_contrib) / port_var

risk_contrib_df = pd.DataFrame({
    "weight": weights.values,
    "risk_contribution": risk_contrib.flatten()
}, index=weights.index)

# =========================================================
# 14) WTI FACTOR EXPOSURE
# =========================================================

wti_beta = simple_beta(port_ret, wti_ret)

# shock day masks
wti_std = wti_ret.std(ddof=1)

wti_half_up = wti_ret > 0.5 * wti_std
wti_half_down = wti_ret < -0.5 * wti_std
wti_one_up = wti_ret > 1.0 * wti_std
wti_one_down = wti_ret < -1.0 * wti_std

# average daily portfolio return during those shock days
avg_port_ret_half_up = port_ret[wti_half_up].mean()
avg_port_ret_half_down = port_ret[wti_half_down].mean()
avg_port_ret_one_up = port_ret[wti_one_up].mean()
avg_port_ret_one_down = port_ret[wti_one_down].mean()

# factor beta during those shock days
wti_beta_half_up = simple_beta(port_ret[wti_half_up], wti_ret[wti_half_up])
wti_beta_half_down = simple_beta(port_ret[wti_half_down], wti_ret[wti_half_down])
wti_beta_one_up = simple_beta(port_ret[wti_one_up], wti_ret[wti_one_up])
wti_beta_one_down = simple_beta(port_ret[wti_one_down], wti_ret[wti_one_down])

# combined absolute shock beta if you want the broad shock regime as well
wti_beta_abs_half = simple_beta(port_ret[abs(wti_ret) > 0.5 * wti_std], wti_ret[abs(wti_ret) > 0.5 * wti_std])
wti_beta_abs_one = simple_beta(port_ret[abs(wti_ret) > 1.0 * wti_std], wti_ret[abs(wti_ret) > 1.0 * wti_std])

# =========================================================
# 15) MONTHLY SMOOTHED SERIES FROM DAILY OUTPUTS
# =========================================================
# Monthly smoothing = averaging daily rolling series inside each month.
# This is NOT recalculating beta/vol/VaR on monthly returns.

monthly_smoothed_vol = roll_vol.resample("ME").mean()
monthly_smoothed_var = roll_var.resample("ME").mean()
monthly_smoothed_beta = roll_beta.resample("ME").mean()
monthly_smoothed_down_beta = roll_down_beta.resample("ME").mean()
monthly_smoothed_up_beta = roll_up_beta.resample("ME").mean()

# also useful to keep monthly portfolio index for display
monthly_port_index = port_index.resample("ME").last()
monthly_drawdown = drawdown.resample("ME").last()

# =========================================================
# 16) SUMMARY TABLES
# =========================================================

summary_df = pd.DataFrame({
    "metric": [
        "portfolio_value_zar",
        "daily_ann_vol_latest",
        "daily_95pct_var_latest",
        "daily_beta_latest",
        "daily_down_beta_latest",
        "daily_up_beta_latest",
        "max_drawdown",
        "cagr",
        "sharpe_rf_4pct",
        "calmar",
        "wti_beta",
        "wti_beta_abs_0_5sd",
        "wti_beta_abs_1sd",
        "avg_port_ret_wti_plus_0_5sd",
        "avg_port_ret_wti_minus_0_5sd",
        "avg_port_ret_wti_plus_1sd",
        "avg_port_ret_wti_minus_1sd",
        "wti_beta_plus_0_5sd",
        "wti_beta_minus_0_5sd",
        "wti_beta_plus_1sd",
        "wti_beta_minus_1sd",
        "n_obs_wti_plus_0_5sd",
        "n_obs_wti_minus_0_5sd",
        "n_obs_wti_plus_1sd",
        "n_obs_wti_minus_1sd",
    ],
    "value": [
        float(sum(NOTIONALS.values())),
        roll_vol.dropna().iloc[-1] if not roll_vol.dropna().empty else np.nan,
        roll_var.dropna().iloc[-1] if not roll_var.dropna().empty else np.nan,
        roll_beta.dropna().iloc[-1] if not roll_beta.dropna().empty else np.nan,
        roll_down_beta.dropna().iloc[-1] if not roll_down_beta.dropna().empty else np.nan,
        roll_up_beta.dropna().iloc[-1] if not roll_up_beta.dropna().empty else np.nan,
        max_drawdown,
        cagr,
        sharpe,
        calmar,
        wti_beta,
        wti_beta_abs_half,
        wti_beta_abs_one,
        avg_port_ret_half_up,
        avg_port_ret_half_down,
        avg_port_ret_one_up,
        avg_port_ret_one_down,
        wti_beta_half_up,
        wti_beta_half_down,
        wti_beta_one_up,
        wti_beta_one_down,
        int(wti_half_up.sum()),
        int(wti_half_down.sum()),
        int(wti_one_up.sum()),
        int(wti_one_down.sum()),
    ]
})

shock_table_df = pd.DataFrame({
    "shock_bucket": [
        "WTI > +0.5σ",
        "WTI < -0.5σ",
        "WTI > +1.0σ",
        "WTI < -1.0σ",
        "|WTI| > 0.5σ",
        "|WTI| > 1.0σ",
    ],
    "n_obs": [
        int(wti_half_up.sum()),
        int(wti_half_down.sum()),
        int(wti_one_up.sum()),
        int(wti_one_down.sum()),
        int((abs(wti_ret) > 0.5 * wti_std).sum()),
        int((abs(wti_ret) > 1.0 * wti_std).sum()),
    ],
    "avg_portfolio_return": [
        avg_port_ret_half_up,
        avg_port_ret_half_down,
        avg_port_ret_one_up,
        avg_port_ret_one_down,
        port_ret[abs(wti_ret) > 0.5 * wti_std].mean(),
        port_ret[abs(wti_ret) > 1.0 * wti_std].mean(),
    ],
    "factor_beta_to_wti": [
        wti_beta_half_up,
        wti_beta_half_down,
        wti_beta_one_up,
        wti_beta_one_down,
        wti_beta_abs_half,
        wti_beta_abs_one,
    ]
})

monthly_smoothed_table_df = pd.DataFrame({
    "monthly_smoothed_vol": monthly_smoothed_vol,
    "monthly_smoothed_var": monthly_smoothed_var,
    "monthly_smoothed_beta": monthly_smoothed_beta,
    "monthly_smoothed_down_beta": monthly_smoothed_down_beta,
    "monthly_smoothed_up_beta": monthly_smoothed_up_beta,
})

# =========================================================
# 17) PRINT EVERYTHING EXPLICITLY
# =========================================================

pd.set_option("display.max_rows", 500)
pd.set_option("display.max_columns", 50)
pd.set_option("display.width", 200)

print("\n=== SUMMARY TABLE ===")
print(summary_df.to_string(index=False))

print("\n=== ASSET-LEVEL DIAGNOSTICS ===")
print(asset_stats_df.round(6).to_string())

print("\n=== RISK CONTRIBUTION TABLE ===")
print(risk_contrib_df.round(6).to_string())

print("\n=== WTI SHOCK TABLE ===")
print(shock_table_df.round(6).to_string(index=False))

print("\n=== MONTHLY SMOOTHED TABLE (TAIL 24 MONTHS) ===")
print(monthly_smoothed_table_df.tail(24).round(6).to_string())

# =========================================================
# 18) FULL VISUAL PACK
# =========================================================

plt.figure(figsize=(14, 6))
plt.plot(port_index, label="Portfolio Index")
plt.title("Portfolio Growth (Daily Cumulative Index)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

plt.figure(figsize=(14, 6))
plt.plot(drawdown, label="Daily Drawdown")
plt.title("Portfolio Drawdown (Daily)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

plt.figure(figsize=(14, 6))
plt.plot(roll_vol, label="180D Rolling Annualized Volatility")
plt.title("Daily Rolling Volatility")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

plt.figure(figsize=(14, 6))
plt.plot(roll_var, label="180D Rolling 95% Historical VaR")
plt.title("Daily Rolling Historical VaR")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

plt.figure(figsize=(14, 6))
plt.plot(roll_beta, label="180D Rolling Beta")
plt.plot(roll_down_beta, label="180D Rolling Down Beta")
plt.plot(roll_up_beta, label="180D Rolling Up Beta")
plt.title("Daily Rolling Beta vs MSCI World")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

plt.figure(figsize=(14, 6))
plt.plot(monthly_port_index, label="Monthly Portfolio Index")
plt.title("Portfolio Cumulative Index (Monthly)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

plt.figure(figsize=(14, 6))
plt.plot(monthly_smoothed_vol, label="Monthly Smoothed Volatility")
plt.plot(monthly_smoothed_var, label="Monthly Smoothed VaR")
plt.title("Monthly Smoothed Risk Metrics")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

plt.figure(figsize=(14, 6))
plt.plot(monthly_smoothed_beta, label="Monthly Smoothed Beta")
plt.plot(monthly_smoothed_down_beta, label="Monthly Smoothed Down Beta")
plt.plot(monthly_smoothed_up_beta, label="Monthly Smoothed Up Beta")
plt.title("Monthly Smoothed Betas vs MSCI World")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

plt.figure(figsize=(14, 6))
plt.bar(risk_contrib_df.index, risk_contrib_df["risk_contribution"])
plt.title("Risk Contribution by Asset")
plt.xticks(rotation=45)
plt.grid(True, axis="y", alpha=0.3)
plt.show()

plt.figure(figsize=(14, 6))
plt.bar(asset_stats_df.index, asset_stats_df["daily_ann_vol"])
plt.title("Asset-Level Annualized Volatility")
plt.xticks(rotation=45)
plt.grid(True, axis="y", alpha=0.3)
plt.show()

plt.figure(figsize=(14, 6))
plt.bar(asset_stats_df.index, asset_stats_df["beta"], label="Beta")
plt.title("Asset-Level Beta vs MSCI World")
plt.xticks(rotation=45)
plt.grid(True, axis="y", alpha=0.3)
plt.show()

plt.figure(figsize=(14, 6))
plt.hist(port_ret, bins=60, alpha=0.7)
plt.title("Distribution of Daily Portfolio Returns")
plt.grid(True, alpha=0.3)
plt.show()

plt.figure(figsize=(14, 6))
plt.scatter(wti_ret.align(port_ret, join="inner")[0], port_ret.align(wti_ret, join="inner")[0], alpha=0.35)
plt.title("Portfolio Return vs WTI Return")
plt.xlabel("WTI Daily Return")
plt.ylabel("Portfolio Daily Return")
plt.grid(True, alpha=0.3)
plt.show()

# WTI shock average daily portfolio returns visualization
shock_avg_plot = pd.Series({
    "WTI > +0.5σ": avg_port_ret_half_up,
    "WTI < -0.5σ": avg_port_ret_half_down,
    "WTI > +1σ": avg_port_ret_one_up,
    "WTI < -1σ": avg_port_ret_one_down,
})
plt.figure(figsize=(10, 6))
plt.bar(shock_avg_plot.index, shock_avg_plot.values)
plt.title("Average Daily Portfolio Return During WTI Shock Days")
plt.grid(True, axis="y", alpha=0.3)
plt.show()

shock_beta_plot = pd.Series({
    "WTI > +0.5σ": wti_beta_half_up,
    "WTI < -0.5σ": wti_beta_half_down,
    "WTI > +1σ": wti_beta_one_up,
    "WTI < -1σ": wti_beta_one_down,
    "|WTI| > 0.5σ": wti_beta_abs_half,
    "|WTI| > 1σ": wti_beta_abs_one,
})
plt.figure(figsize=(10, 6))
plt.bar(shock_beta_plot.index, shock_beta_plot.values)
plt.title("Portfolio Factor Beta to WTI in Shock Buckets")
plt.grid(True, axis="y", alpha=0.3)
plt.xticks(rotation=20)
plt.show()
```
