```python
# =====================================================
# FULL PROJECT (DO NOT REMOVE)
# Staples vs Discretionary
# Z-Score + ACF + Rolling AR(1) + Rolling AR(12) + Rolling Chow Evolution
# =====================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf
from scipy.stats import f

# -----------------------------------------------------
# Professional Academic Styling
# -----------------------------------------------------
plt.style.use("seaborn-v0_8-darkgrid")
plt.rcParams.update({
    "font.family": "serif",
    "font.serif": ["Times New Roman", "DejaVu Serif"],
    "axes.labelsize": 12,
    "axes.titlesize": 14,
    "xtick.labelsize": 10,
    "ytick.labelsize": 10,
})

# -----------------------------------------------------
# Helpers
# -----------------------------------------------------
def _clean_price_col(s: pd.Series) -> pd.Series:
    return s.astype(str).str.replace(",", "", regex=False).astype(float)

def build_ar_design(y: np.ndarray, p: int):
    """
    For AR(p): y_t = a + b1 y_{t-1} + ... + bp y_{t-p} + e_t
    Returns (Y, X) aligned so Y has length n-p and X has shape (n-p, p+1).
    """
    n = len(y)
    if n <= p:
        raise ValueError(f"Not enough observations ({n}) for AR({p}).")
    Y = y[p:]
    X_lags = np.column_stack([y[p - k: n - k] for k in range(1, p + 1)])  # y_{t-1}..y_{t-p}
    X = sm.add_constant(X_lags)
    return Y, X

def ar_rss(series: np.ndarray, p: int) -> float:
    Y, X = build_ar_design(series, p)
    model = sm.OLS(Y, X).fit()
    resid = model.resid
    return float(np.sum(resid ** 2))

def rolling_ar_betas(returns: pd.Series, p: int, window: int) -> pd.DataFrame:
    """
    Rolling AR(p) coefficients over 'window' observations.
    Returns DataFrame with columns: const, lag1..lagp
    """
    vals = returns.values
    idx = returns.index
    rows = []
    out_index = []
    for i in range(window, len(vals) + 1):
        sample = vals[i - window:i]
        if np.any(np.isnan(sample)) or len(sample) <= p:
            continue
        Y, X = build_ar_design(sample, p)
        fit = sm.OLS(Y, X).fit()
        params = fit.params  # [const, b1..bp]
        rows.append(params)
        out_index.append(idx[i - 1])  # stamp at window end
    cols = ["const"] + [f"lag{k}" for k in range(1, p + 1)]
    return pd.DataFrame(rows, index=pd.DatetimeIndex(out_index), columns=cols)

def plot_last_label(ax, x, y, fmt, offset=(8, 6), fontsize=11):
    ax.scatter(x, y, s=60)
    ax.annotate(fmt.format(y), (x, y), textcoords="offset points",
                xytext=offset, fontsize=fontsize)

def rolling_chow_F(series: np.ndarray, dates: pd.DatetimeIndex, p: int, min_window: int):
    """
    Compute Chow F-stat at each candidate break date.
    Uses AR(p). k = p+1 parameters.
    Returns pd.Series(F_t) indexed by candidate break date and a 5% critical value line.
    """
    k = p + 1
    n = len(series)
    if n <= 2 * min_window:
        raise ValueError(f"Not enough observations for min_window={min_window}. Have {n}.")

    rss_pooled = ar_rss(series, p)

    F_vals = []
    F_dates = []
    df2_vals = []

    # break at t means left: [0..t-1], right: [t..n-1]
    for t in range(min_window, n - min_window + 1):
        left = series[:t]
        right = series[t:]

        # Need each side to have > p observations for AR(p)
        if len(left) <= p or len(right) <= p:
            continue

        rss1 = ar_rss(left, p)
        rss2 = ar_rss(right, p)

        n1 = len(left) - p
        n2 = len(right) - p

        denom_df = (n1 + n2 - 2 * k)
        if denom_df <= 0:
            continue

        numerator = (rss_pooled - (rss1 + rss2)) / k
        denominator = (rss1 + rss2) / denom_df

        F_t = numerator / denominator
        F_vals.append(F_t)
        F_dates.append(dates[t - 1])  # stamp at last date of left segment
        df2_vals.append(denom_df)

    F_series = pd.Series(F_vals, index=pd.DatetimeIndex(F_dates)).replace([np.inf, -np.inf], np.nan).dropna()

    # Use conservative df2 for a single reference critical line
    df1 = k
    df2_min = int(np.min(df2_vals)) if len(df2_vals) else (n - 2 * k)
    crit_5 = f.ppf(0.95, df1, df2_min)

    return F_series, crit_5

# -----------------------------------------------------
# Load Data
# -----------------------------------------------------
kxi = pd.read_csv("KXI Historical Data.csv")
rxi = pd.read_csv("RXI ETF Stock Price History.csv")

kxi["Date"] = pd.to_datetime(kxi["Date"])
rxi["Date"] = pd.to_datetime(rxi["Date"])

kxi = kxi.sort_values("Date")
rxi = rxi.sort_values("Date")

kxi["Price"] = _clean_price_col(kxi["Price"])
rxi["Price"] = _clean_price_col(rxi["Price"])

df = pd.merge(
    kxi[["Date", "Price"]],
    rxi[["Date", "Price"]],
    on="Date",
    how="inner",
    suffixes=("_KXI", "_RXI")
).set_index("Date")

# -----------------------------------------------------
# Weekly Log Spread + Returns
# -----------------------------------------------------
weekly = df.resample("W-FRI").last().dropna()
weekly["Log_Spread"] = np.log(weekly["Price_KXI"]) - np.log(weekly["Price_RXI"])
weekly["Spread_Return"] = weekly["Log_Spread"].diff()
weekly = weekly.dropna()

# =====================================================
# 1) Rolling 52-Week Z-Score (of Log Spread)
# =====================================================
lookback = 52
mu = weekly["Log_Spread"].rolling(lookback).mean()
sd = weekly["Log_Spread"].rolling(lookback).std()
weekly["ZScore"] = (weekly["Log_Spread"] - mu) / sd

fig1, ax1 = plt.subplots(figsize=(12, 6))
ax1.plot(weekly.index, weekly["ZScore"], linewidth=1.6)
ax1.axhline(0, linestyle="--", linewidth=1)
ax1.axhline(2, linestyle="--", linewidth=1)
ax1.axhline(-2, linestyle="--", linewidth=1)

z_last_date = weekly.index[-1]
z_last_val = weekly["ZScore"].iloc[-1]
plot_last_label(ax1, z_last_date, z_last_val, "{:.2f}")

ax1.set_title("Rolling 52-Week Z-Score\nStaples vs Discretionary (Log Spread)")
ax1.set_xlabel("Date")
ax1.set_ylabel("Z-Score")
plt.tight_layout()
plt.show()

# =====================================================
# 2) ACF (1–52 lags) of Weekly Spread Returns
# =====================================================
fig2, ax2 = plt.subplots(figsize=(12, 6))
plot_acf(weekly["Spread_Return"], lags=52, alpha=0.05, ax=ax2)
ax2.set_title("ACF — Weekly Spread Returns (1–52 Lags)")
ax2.set_xlabel("Lag (Weeks)")
ax2.set_ylabel("Autocorrelation")
plt.tight_layout()
plt.show()

# =====================================================
# 3) Rolling AR(1) Beta (52-week window)
# =====================================================
roll_window = 52
ar1 = rolling_ar_betas(weekly["Spread_Return"], p=1, window=roll_window)
beta1 = ar1["lag1"].dropna()

fig3, ax3 = plt.subplots(figsize=(12, 6))
ax3.plot(beta1.index, beta1, linewidth=1.6)
ax3.axhline(0, linestyle="--", linewidth=1)

b1_last_date = beta1.index[-1]
b1_last_val = beta1.iloc[-1]
plot_last_label(ax3, b1_last_date, b1_last_val, "{:.3f}")

ax3.set_title("Rolling AR(1) Coefficient (Lag 1)\n52-Week Rolling Window")
ax3.set_xlabel("Date")
ax3.set_ylabel("β₁")
plt.tight_layout()
plt.show()

# =====================================================
# 4) Rolling AR(12) Persistence (sum of betas, lags 1–12)
# =====================================================
ar12 = rolling_ar_betas(weekly["Spread_Return"], p=12, window=roll_window)
lag_cols = [f"lag{k}" for k in range(1, 13)]
ar12_sum = ar12[lag_cols].sum(axis=1).dropna()  # persistence proxy

fig4, ax4 = plt.subplots(figsize=(12, 6))
ax4.plot(ar12_sum.index, ar12_sum, linewidth=1.6)
ax4.axhline(0, linestyle="--", linewidth=1)

s_last_date = ar12_sum.index[-1]
s_last_val = ar12_sum.iloc[-1]
plot_last_label(ax4, s_last_date, s_last_val, "{:.3f}")

ax4.set_title("Rolling AR(12) Persistence\nSum of β₁..β₁₂ (52-Week Rolling Window)")
ax4.set_xlabel("Date")
ax4.set_ylabel("Σβ (lags 1–12)")
plt.tight_layout()
plt.show()

# =====================================================
# 5) Rolling Chow F Evolution — AR(1)
# =====================================================
r = weekly["Spread_Return"].values
dates = weekly.index

F_ar1, crit_ar1 = rolling_chow_F(r, dates, p=1, min_window=52)

fig5, ax5 = plt.subplots(figsize=(12, 6))
ax5.plot(F_ar1.index, F_ar1, linewidth=1.6)
ax5.axhline(crit_ar1, linestyle="--", linewidth=1)

F1_last_date = F_ar1.index[-1]
F1_last_val = F_ar1.iloc[-1]
plot_last_label(ax5, F1_last_date, F1_last_val, "{:.2f}")

ax5.set_title("Rolling Chow F-Statistic Evolution — AR(1)")
ax5.set_xlabel("Candidate Break Date")
ax5.set_ylabel("Chow F")
plt.tight_layout()
plt.show()

# =====================================================
# 6) Rolling Chow F Evolution — AR(12)
# (needs more data on each side; use a larger min window)
# =====================================================
F_ar12, crit_ar12 = rolling_chow_F(r, dates, p=12, min_window=104)

fig6, ax6 = plt.subplots(figsize=(12, 6))
ax6.plot(F_ar12.index, F_ar12, linewidth=1.6)
ax6.axhline(crit_ar12, linestyle="--", linewidth=1)

F12_last_date = F_ar12.index[-1]
F12_last_val = F_ar12.iloc[-1]
plot_last_label(ax6, F12_last_date, F12_last_val, "{:.2f}")

ax6.set_title("Rolling Chow F-Statistic Evolution — AR(12)")
ax6.set_xlabel("Candidate Break Date")
ax6.set_ylabel("Chow F")
plt.tight_layout()
plt.show()

# -----------------------------------------------------
# Print latest values
# -----------------------------------------------------
print(f"Latest Z-Score: {z_last_val:.4f}")
print(f"Latest Rolling AR(1) β1: {b1_last_val:.6f}")
print(f"Latest Rolling AR(12) Σβ(1..12): {s_last_val:.6f}")
print(f"Latest Chow F (AR1): {F1_last_val:.6f} | 5% crit: {crit_ar1:.6f}")
print(f"Latest Chow F (AR12): {F12_last_val:.6f} | 5% crit: {crit_ar12:.6f}")

```
