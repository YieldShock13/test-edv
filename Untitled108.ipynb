```python
# ============================================================
# COLAB-READY  FULL FRESH SCRIPT (SEMI-ANNUAL  ROBUST  LABELED)
# What you get (semi-annual, Jun/Dec period-end):
#   1) Rolling Z of (ALSI  S&P 500) on returns
#   2) Rolling corr: VIX(level) vs ALSI & sectors (returns)
#   3) Rolling corr: Precious Metals (returns) vs ALSI & sectors (returns)
#   4) VIXEquity Z-spread: z(VIX level)  z(Return)
#   5) Sovereign spread SA10YUS10Y (levels) + rolling Z
#   6) Z-spread: z(SA10Y_bps)  z(ALSI_return)
#   7) USD/ZAR rolling Z on returns
#
# Features:
#    Exact filenames as you uploaded (with keyword fallback)
#    Pairwise inner-join alignment to guarantee overlap
#    Adaptive rolling window/min_periods so shorter histories still render
#    Right-edge labels & watermark on every chart
#    PNGs + latest CSVs saved to CURRENT DIR
# ============================================================

import os, re, glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ------------------- PARAMS -------------------
RESAMPLE_FREQ   = "2Q-DEC"     # semi-annual (Jun/Dec)
WINDOW_PERIODS  = 8            # rolling window length in semi-annual periods (~4y)
MIN_PERIODS     = 4
WATERMARK       = "orca's capital strategies"
OUT_DIR         = os.getcwd()
os.makedirs(OUT_DIR, exist_ok=True)

# Exact basenames (as provided)
EXPECTED = {
    # JSE
    "All Share": "FTSE_JSE All Share Historical Data.csv",
    "Industrial 25": "FTSE_JSE Industrial 25 Historical Data.csv",
    "Financial 15":  "FTSE_JSE Financial 15 Historical Data.csv",
    "REITs":         "FTSE JSE Real Estate Investment Trusts Historical Data.csv",
    "Resource 10":   "FTSE_JSE Resource 10 Historical Data.csv",
    # Other assets
    "S&P 500":       "S&P 500 Historical Data (5).csv",
    "VIX":           "CBOE Volatility Index Historical Data (3).csv",
    "PreciousMetals":"GSCI Precious Metal Historical Data.csv",
    "SA10Y":         "South Africa 10-Year Bond Yield Historical Data.csv",
    "US10Y":         "United States 10-Year Bond Yield Historical Data (9).csv",
    "USDZAR":        "USD_ZAR Historical Data.csv",
}
SEARCH_DIRS = [os.getcwd(), "/mnt/data", "/content"]

# ------------------- DISCOVERY & IO -------------------
def discover_paths(expected_map, search_dirs):
    found = {k: None for k in expected_map}
    # exact
    for label, base in expected_map.items():
        for d in search_dirs:
            p = os.path.join(d, base)
            if os.path.exists(p):
                found[label] = p
                break
    # keyword fallback (if any missing)
    if any(v is None for v in found.values()):
        kw = {
            "All Share": ["all share"],
            "Industrial 25": ["industrial 25"],
            "Financial 15": ["financial 15"],
            "REITs": ["real estate investment trusts","reits"],
            "Resource 10": ["resource 10"],
            "S&P 500": ["s&p 500","sp500","sp 500","s&p"],
            "VIX": ["volatility index","vix"],
            "PreciousMetals": ["precious","metal","gsci"],
            "SA10Y": ["south africa 10-year","south africa 10 year","za 10-year","za 10 year"],
            "US10Y": ["united states 10-year","us 10-year","us 10 year"],
            "USDZAR": ["usd_zar","usdzar","usd-zar","usd/zar"],
        }
        pool = []
        for d in search_dirs:
            pool += glob.glob(os.path.join(d, "*.csv"))
        for label, path in found.items():
            if path is None:
                for p in pool:
                    nm = os.path.basename(p).lower()
                    if any(token in nm for token in kw[label]):
                        found[label] = p
                        break
    missing = [k for k,v in found.items() if v is None]
    if missing:
        raise FileNotFoundError(f"Missing CSV(s) for: {missing}. Check names/locations.")
    return found

def _to_float(x):
    if isinstance(x, str):
        x = re.sub(r"[^0-9.\-]", "", x.replace(",","").replace(" ",""))
    try: return float(x)
    except: return np.nan

def read_investing_csv(path: str) -> pd.Series:
    df = pd.read_csv(path)
    df.columns = [c.strip() for c in df.columns]
    date_col = next((c for c in df.columns if c.lower()=="date"), None)
    if date_col is None:
        raise ValueError(f"No Date column in {path}")
    df[date_col] = pd.to_datetime(df[date_col], errors="coerce")

    price_candidates = ["Price","Close","Last","Adj Close"]
    price_col = next((c for c in price_candidates if c in df.columns), None)
    if price_col is None:
        price_col = next((c for c in df.columns if c.lower() in [p.lower() for p in price_candidates]), None)
    if price_col is None:
        raise ValueError(f"No price/close column in {path}")

    df[price_col] = df[price_col].apply(_to_float)
    df = df.dropna(subset=[date_col, price_col]).sort_values(date_col)

    # Semi-annual alignment (end-of-period)
    s = df.set_index(date_col)[price_col].resample(RESAMPLE_FREQ).last().dropna()
    return s.rename("price")

# ------------------- TRANSFORMS & STATS -------------------
def sa_log_returns(price: pd.Series) -> pd.Series:
    return np.log(price).diff()

def _adaptive_window_len(series: pd.Series, default_w=WINDOW_PERIODS, default_min=MIN_PERIODS):
    n = int(series.dropna().shape[0])
    if n <= 2:
        return 2, 2
    w = min(default_w, n)
    m = min(default_min, max(2, w//2))
    return w, m

def _pair_align(a: pd.Series, b: pd.Series):
    """Inner-join alignment (overlap only)."""
    df = pd.concat([a, b], axis=1, join='inner').dropna()
    if df.empty:
        return a.reindex([]), b.reindex([])
    return df.iloc[:,0], df.iloc[:,1]

def sa_rolling_z(series: pd.Series) -> pd.Series:
    s = series.dropna()
    if s.empty: return s
    w, m = _adaptive_window_len(s)
    mu = s.rolling(w, min_periods=m).mean()
    sd = s.rolling(w, min_periods=m).std(ddof=0)
    z  = (s - mu) / sd
    return z.dropna()

def sa_rolling_z_of_spread(a: pd.Series, b: pd.Series) -> pd.Series:
    a2, b2 = _pair_align(a, b)
    if a2.empty: return a2
    spread = a2 - b2
    return sa_rolling_z(spread)

def sa_rolling_corr(a: pd.Series, b: pd.Series) -> pd.Series:
    a2, b2 = _pair_align(a, b)
    if a2.empty: return a2
    df = pd.concat([a2, b2], axis=1).dropna()
    w, m = _adaptive_window_len(df.iloc[:,0])
    if w < 2: return df.iloc[:,0].reindex([])
    return df.iloc[:,0].rolling(w, min_periods=m).corr(df.iloc[:,1]).dropna()

# ------------------- PLOTTING HELPERS -------------------
def add_sigma_bands(ax):
    for lvl in [0,1,2,3,-1,-2,-3]:
        ax.axhline(y=lvl, linestyle="--" if lvl else "-", linewidth=1, alpha=0.6)

def add_zero_line(ax):
    ax.axhline(y=0, linestyle="--", linewidth=1, alpha=0.5)

def add_watermark(ax, text=WATERMARK):
    ax.text(0.995, 0.02, text, transform=ax.transAxes, ha="right", va="bottom", alpha=0.5, fontsize=9)

def label_last_series(ax, s, label):
    s = s.dropna()
    if s.empty: return
    ld = s.index.max(); val = s.loc[ld]
    ax.annotate(f"{label}: {val:.2f}", xy=(ld, val), xytext=(10, 0),
                textcoords="offset points", va="center", fontsize=9,
                bbox=dict(boxstyle="round,pad=0.15", fc="white", ec="none", alpha=0.7))

def label_last_df(ax, df):
    if df.empty: return
    ld = df.index.max()
    vals = df.loc[ld].dropna().sort_values()
    if vals.empty: return
    n = len(vals); nudges = np.linspace(-20, 20, n) if n > 1 else [0]
    for (name, v), dy in zip(vals.items(), nudges):
        ax.annotate(f"{name}: {v:.2f}", xy=(ld, v), xytext=(10, dy),
                    textcoords="offset points", va="center", fontsize=9,
                    bbox=dict(boxstyle="round,pad=0.15", fc="white", ec="none", alpha=0.7))

# --- Robust saver (handles Series scalar & DataFrame row correctly) ---
def save_latest_csv(obj, fname: str, colname: str = "latest"):
    path = os.path.join(OUT_DIR, fname)

    if isinstance(obj, pd.Series):
        s = obj.dropna()
        if s.empty:
            return
        last_idx = s.index.max()
        last_val = s.loc[last_idx]
        out = pd.DataFrame({colname: [last_val]}, index=[last_idx])
        out.to_csv(path)
        return

    if isinstance(obj, pd.DataFrame):
        df = obj.dropna(how="all")
        if df.empty:
            return
        last_idx = df.index.max()
        row = df.loc[last_idx].dropna()
        out = pd.DataFrame(row).T
        out.index = [last_idx]
        out.to_csv(path)
        return

    # Fallback for scalars
    try:
        out = pd.DataFrame({colname: [float(obj)]})
        out.to_csv(path, index=False)
    except Exception:
        pass

# ------------------- LOAD (SEMI-ANNUAL ALIGNED) -------------------
paths = discover_paths(EXPECTED, SEARCH_DIRS)

alshi = read_investing_csv(paths["All Share"])
sectors = {
    "Industrial 25": read_investing_csv(paths["Industrial 25"]),
    "Financial 15":  read_investing_csv(paths["Financial 15"]),
    "REITs":         read_investing_csv(paths["REITs"]),
    "Resource 10":   read_investing_csv(paths["Resource 10"]),
}
spx    = read_investing_csv(paths["S&P 500"])
vix    = read_investing_csv(paths["VIX"])
pm     = read_investing_csv(paths["PreciousMetals"])
sa10y  = read_investing_csv(paths["SA10Y"])   # level, %
us10y  = read_investing_csv(paths["US10Y"])   # level, %
usdzar = read_investing_csv(paths["USDZAR"])

# Semi-annual returns
alshi_r   = sa_log_returns(alshi)
sectors_r = {k: sa_log_returns(v) for k, v in sectors.items()}
spx_r     = sa_log_returns(spx)
pm_r      = sa_log_returns(pm)
usdzar_r  = sa_log_returns(usdzar)

# SA10Y in bps
sa10y_d_bps = sa10y.diff() * 100.0

# ============================================================
# 1) Z(ALSI  S&P500) on returns
# ============================================================
z_jse_minus_spx = sa_rolling_z_of_spread(alshi_r, spx_r)
if not z_jse_minus_spx.dropna().empty:
    fig, ax = plt.subplots(figsize=(12,6))
    ax.plot(z_jse_minus_spx.index, z_jse_minus_spx.values, label="Z( ALSI  S&P500 )")
    add_sigma_bands(ax); add_watermark(ax); label_last_series(ax, z_jse_minus_spx, "ALSIS&P500")
    ax.set_title(f"Rolling Z (SEMI-ANNUAL): ALSI vs S&P 500 (returns)")
    ax.set_xlabel("Date"); ax.set_ylabel("Z-score"); ax.grid(alpha=0.3); fig.tight_layout()
    fig.savefig(os.path.join(OUT_DIR, "A_Z_ALSI_minus_SPX_SA.png"), dpi=150); plt.show()
    save_latest_csv(z_jse_minus_spx, "A_latest_Z_ALSI_minus_SPX_SA.csv")

# ============================================================
# 2) Rolling corr: VIX(level) vs ALSI & sectors (returns)
# ============================================================
corr_vix = {}
c_vix_alsi = sa_rolling_corr(vix, alshi_r)
if not c_vix_alsi.dropna().empty:
    corr_vix["All Share"] = c_vix_alsi
for name, r in sectors_r.items():
    c = sa_rolling_corr(vix, r)
    if not c.dropna().empty:
        corr_vix[name] = c
corr_vix_df = pd.DataFrame(corr_vix).dropna(how="all")

if not corr_vix_df.empty:
    fig, ax = plt.subplots(figsize=(14,7))
    for col in corr_vix_df.columns:
        ax.plot(corr_vix_df.index, corr_vix_df[col], label=col)
    add_zero_line(ax); add_watermark(ax); label_last_df(ax, corr_vix_df)
    ax.set_title("Rolling Correlation (SEMI-ANNUAL): VIX (level) vs JSE (returns)")
    ax.set_xlabel("Date"); ax.set_ylabel("Correlation")
    ax.legend(ncol=2, frameon=False); ax.grid(alpha=0.3); fig.tight_layout()
    fig.savefig(os.path.join(OUT_DIR, "B_Corr_VIX_vs_JSE_AllSectors_MAIN_SA.png"), dpi=150); plt.show()
    for col in corr_vix_df.columns:
        fig, ax = plt.subplots(figsize=(12,6))
        ax.plot(corr_vix_df.index, corr_vix_df[col], label=col)
        add_zero_line(ax); add_watermark(ax); label_last_series(ax, corr_vix_df[col], col)
        ax.set_title(f"Rolling Corr (SEMI-ANNUAL): VIX vs {col}")
        ax.set_xlabel("Date"); ax.set_ylabel("Correlation"); ax.grid(alpha=0.3); fig.tight_layout()
        fig.savefig(os.path.join(OUT_DIR, f"B_Corr_VIX_vs_{col.replace(' ','_')}_SA.png"), dpi=150); plt.show()
    save_latest_csv(corr_vix_df, "B_latest_Corr_VIX_vs_JSE_SA.csv")

# ============================================================
# 3) Rolling corr: Precious Metals (returns) vs ALSI & sectors (returns)
# ============================================================
corr_pm = {}
c_pm_alsi = sa_rolling_corr(pm_r, alshi_r)
if not c_pm_alsi.dropna().empty:
    corr_pm["All Share"] = c_pm_alsi
for name, r in sectors_r.items():
    c = sa_rolling_corr(pm_r, r)
    if not c.dropna().empty:
        corr_pm[name] = c
corr_pm_df = pd.DataFrame(corr_pm).dropna(how="all")

if not corr_pm_df.empty:
    fig, ax = plt.subplots(figsize=(14,7))
    for col in corr_pm_df.columns:
        ax.plot(corr_pm_df.index, corr_pm_df[col], label=col)
    add_zero_line(ax); add_watermark(ax); label_last_df(ax, corr_pm_df)
    ax.set_title("Rolling Correlation (SEMI-ANNUAL): Precious Metals vs JSE (returns)")
    ax.set_xlabel("Date"); ax.set_ylabel("Correlation")
    ax.legend(ncol=2, frameon=False); ax.grid(alpha=0.3); fig.tight_layout()
    fig.savefig(os.path.join(OUT_DIR, "C_Corr_PM_vs_JSE_AllSectors_MAIN_SA.png"), dpi=150); plt.show()
    for col in corr_pm_df.columns:
        fig, ax = plt.subplots(figsize=(12,6))
        ax.plot(corr_pm_df.index, corr_pm_df[col], label=col)
        add_zero_line(ax); add_watermark(ax); label_last_series(ax, corr_pm_df[col], col)
        ax.set_title(f"Rolling Corr (SEMI-ANNUAL): PM vs {col}")
        ax.set_xlabel("Date"); ax.set_ylabel("Correlation"); ax.grid(alpha=0.3); fig.tight_layout()
        fig.savefig(os.path.join(OUT_DIR, f"C_Corr_PM_vs_{col.replace(' ','_')}_SA.png"), dpi=150); plt.show()
    save_latest_csv(corr_pm_df, "C_latest_Corr_PM_vs_JSE_SA.csv")

# ============================================================
# 4) VIXEquity Z-spread: z(VIX level)  z(Return)
# ============================================================
z_vix = sa_rolling_z(vix)
z_equities = {"All Share": sa_rolling_z(alshi_r)}
for name, r in sectors_r.items():
    z_equities[name] = sa_rolling_z(r)

z_spreads = {}
for k, z_eq in z_equities.items():
    if z_vix.dropna().empty or z_eq.dropna().empty:
        continue
    idx = z_vix.index.intersection(z_eq.index)
    if len(idx) == 0:
        continue
    z_spreads[k] = (z_vix.reindex(idx) - z_eq.reindex(idx)).dropna()

z_spreads_df = pd.DataFrame(z_spreads).dropna(how="all")

if not z_spreads_df.empty:
    fig, ax = plt.subplots(figsize=(14,7))
    for col in z_spreads_df.columns:
        ax.plot(z_spreads_df.index, z_spreads_df[col], label=col)
    add_sigma_bands(ax); add_watermark(ax); label_last_df(ax, z_spreads_df)
    ax.set_title("VIXEquity Z-spread (SEMI-ANNUAL): z(VIX)  z(Return)")
    ax.set_xlabel("Date"); ax.set_ylabel("Z-spread")
    ax.legend(ncol=2, frameon=False); ax.grid(alpha=0.3); fig.tight_layout()
    fig.savefig(os.path.join(OUT_DIR, "D_Zspread_VIX_minus_Equity_MAIN_SA.png"), dpi=150); plt.show()
    for col in z_spreads_df.columns:
        fig, ax = plt.subplots(figsize=(12,6))
        ax.plot(z_spreads_df.index, z_spreads_df[col], label=col)
        add_sigma_bands(ax); add_watermark(ax); label_last_series(ax, z_spreads_df[col], col)
        ax.set_title(f"VIX{col} Z-spread (SEMI-ANNUAL)")
        ax.set_xlabel("Date"); ax.set_ylabel("Z-spread"); ax.grid(alpha=0.3); fig.tight_layout()
        fig.savefig(os.path.join(OUT_DIR, f"D_Zspread_VIX_minus_{col.replace(' ','_')}_SA.png"), dpi=150); plt.show()
    save_latest_csv(z_spreads_df, "D_latest_Zspread_VIX_minus_Equity_SA.csv")

# ============================================================
# 5A) Sovereign spread (levels) & rolling Z
# ============================================================
sa10y_a, us10y_a = _pair_align(sa10y, us10y)
sovereign_spread = (sa10y_a - us10y_a).dropna()
if not sovereign_spread.empty:
    z_sovereign_spread = sa_rolling_z(sovereign_spread)

    fig, ax = plt.subplots(figsize=(12,6))
    ax.plot(sovereign_spread.index, sovereign_spread.values, label="SA10Y  US10Y (pp)")
    add_zero_line(ax); add_watermark(ax); label_last_series(ax, sovereign_spread, "SA10YUS10Y")
    ax.set_title("Sovereign Spread (SEMI-ANNUAL levels): SA10Y  US10Y")
    ax.set_xlabel("Date"); ax.set_ylabel("pp"); ax.grid(alpha=0.3); fig.tight_layout()
    fig.savefig(os.path.join(OUT_DIR, "E_Spread_SA10Y_minus_US10Y_LEVELS_SA.png"), dpi=150); plt.show()
    save_latest_csv(sovereign_spread, "E_latest_SovereignSpread_LEVELS_SA.csv")

    if not z_sovereign_spread.dropna().empty:
        fig, ax = plt.subplots(figsize=(12,6))
        ax.plot(z_sovereign_spread.index, z_sovereign_spread.values, label="Z( SA10Y  US10Y )")
        add_sigma_bands(ax); add_watermark(ax); label_last_series(ax, z_sovereign_spread, "Z(SA10YUS10Y)")
        ax.set_title("Rolling Z (SEMI-ANNUAL): Sovereign Spread")
        ax.set_xlabel("Date"); ax.set_ylabel("Z-score"); ax.grid(alpha=0.3); fig.tight_layout()
        fig.savefig(os.path.join(OUT_DIR, "E_Z_Spread_SA10Y_minus_US10Y_SA.png"), dpi=150); plt.show()
        save_latest_csv(z_sovereign_spread, "E_latest_Z_SovereignSpread_SA.csv")

# ============================================================
# 5B) Z-spread: z(SA10Y_bps)  z(ALSI_return)
# ============================================================
sa10y_d_bps_al, alshi_r_al = _pair_align(sa10y_d_bps, alshi_r)
if not sa10y_d_bps_al.dropna().empty and not alshi_r_al.dropna().empty:
    z_dsa10y = sa_rolling_z(sa10y_d_bps_al)
    z_alsi   = sa_rolling_z(alshi_r_al)
    idx = z_dsa10y.index.intersection(z_alsi.index)
    z_spread_yield_vs_stock = (z_dsa10y.reindex(idx) - z_alsi.reindex(idx)).dropna()

    if not z_spread_yield_vs_stock.dropna().empty:
        fig, ax = plt.subplots(figsize=(12,6))
        ax.plot(z_spread_yield_vs_stock.index, z_spread_yield_vs_stock.values, label="z(SA10Y_bps)  z(ALSI_ret)")
        add_sigma_bands(ax); add_watermark(ax); label_last_series(ax, z_spread_yield_vs_stock, "Z(SA10Y)Z(ALSI)")
        ax.set_title("Z-spread (SEMI-ANNUAL): SA10Y (bps) vs ALSI Return")
        ax.set_xlabel("Date"); ax.set_ylabel("Z-spread"); ax.grid(alpha=0.3); fig.tight_layout()
        fig.savefig(os.path.join(OUT_DIR, "F_Zspread_dSA10Ybps_vs_ALSI_SA.png"), dpi=150); plt.show()
        save_latest_csv(z_spread_yield_vs_stock, "F_latest_Zspread_dSA10Ybps_vs_ALSI_SA.csv")

# ============================================================
# 6) USD/ZAR rolling Z on returns
# ============================================================
z_usdzar = sa_rolling_z(usdzar_r)
if not z_usdzar.dropna().empty:
    fig, ax = plt.subplots(figsize=(12,6))
    ax.plot(z_usdzar.index, z_usdzar.values, label="Z( USD/ZAR returns )")
    add_sigma_bands(ax); add_watermark(ax); label_last_series(ax, z_usdzar, "USDZAR")
    ax.set_title("Rolling Z (SEMI-ANNUAL): USD/ZAR (returns)")
    ax.set_xlabel("Date"); ax.set_ylabel("Z-score"); ax.grid(alpha=0.3); fig.tight_layout()
    fig.savefig(os.path.join(OUT_DIR, "G_USDZAR_RollingZ_SA.png"), dpi=150); plt.show()
    save_latest_csv(z_usdzar, "G_latest_Z_USDZAR_SA.csv")

print("Done (SEMI-ANNUAL, aligned, adaptive). Files saved in:", OUT_DIR)



```
