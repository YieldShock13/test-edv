```python
# ============================================================
# FULL CONSOLIDATED PROJECT  DAILY 2Y/5Y/10Y BUTTERFLY (YIELD SPACE)
# BACK TO BUTTERFLY TRADE
#
# Inputs: DGS2.csv, DGS5 (3).csv, DGS10 (13).csv (already uploaded)
#
# Pipeline per regime (FULL, 2010-2020, 2020-2026):
# 1) Load + clean yields (robust DATE column detection)
# 2) Daily yield changes in bps
# 3) Rolling beta-hedged butterfly via regression: dy5 ~ c + a*dy2 + b*dy10
# 4) fly_delta_bps = residual; fly_level_bps = cumsum
# 5) Rolling Z on fly_level (shift(1), no lookahead)
# 6) STRATEGY (single position, no ladder):
#    ENTER at |z| >= 2.00
#    EXIT ALL when z crosses 0.00 (full mean reversion, same-side)
#    Both sides, non-overlapping, no time cap
# 7) Outputs: trades CSV, daily CSV, summaries, density plots
# 8) OU MLE on fly_level + half-life
# 9) Rolling OU kappa stability + plot + CSV
# 10) Portfolio:
#     NOTIONAL=80k
#     SLIPPAGE=10 bps per TRADE total (once per completed trade; half on entry, half on exit)
#     FINANCING=1.77% annual applied ONLY while in-position (daily)
#     CAGR, MaxDD, Sharpe (Rf=3%), Calmar, VaR95 + winsorized(95%) metrics
# 11) Black plots + watermark + FINAL POINT LABELED
# ============================================================

import os, re, glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.optimize import minimize
from scipy.stats import gaussian_kde

# -------------------------
# USER INPUTS
# -------------------------
Y2_FILE  = "DGS2.csv"
Y5_FILE  = "DGS5 (3).csv"
Y10_FILE = "DGS10 (13).csv"

# Rolling windows (daily)
ROLL_BETA_WINDOW = 252
ROLL_Z_WINDOW    = 252

# Strategy thresholds (YOUR NEW RULES)
ENTRY_LEVEL = 2.00    # enter at |z| >= 2.0
EXIT_LEVEL  = 0.00    # exit when z crosses 0.0

# Portfolio & costs
NOTIONAL_USD = 80000.0
RF_ANNUAL    = 0.03
SLIPPAGE_BPS_PER_TRADE_TOTAL = 10.0   # once per completed trade (half entry day, half exit day)
FINANCING_ANNUAL = 0.0177             # applied ONLY while in position

# OU rolling stability
ROLL_OU_WINDOW = 1260
ROLL_OU_STEP   = 5

# Regimes
REGIMES = [
    ("FULL", None, None),
    ("2010_2020", "2010-01-01", "2020-12-31"),
    ("2020_2026", "2020-01-01", "2026-12-31"),
]

OUT_DIR = "butterfly_full_outputs"
os.makedirs(OUT_DIR, exist_ok=True)

# -------------------------
# FILE FINDER (robust; no /mnt required in names)
# -------------------------
def find_file(preferred_name: str) -> str:
    if os.path.exists(preferred_name):
        return preferred_name
    for base in [".", "/content", "/mnt/data"]:
        p = os.path.join(base, preferred_name)
        if os.path.exists(p):
            return p
    token = re.sub(r"[^A-Za-z0-9]+", "", preferred_name).lower()
    hits = []
    for base in [".", "/content", "/mnt/data"]:
        for p in glob.glob(os.path.join(base, "*.csv")):
            t = re.sub(r"[^A-Za-z0-9]+", "", os.path.basename(p)).lower()
            if token in t:
                hits.append(p)
    if hits:
        hits = sorted(hits, key=lambda x: (len(os.path.basename(x)), x))
        return hits[0]
    raise FileNotFoundError(f"Could not find file: {preferred_name}")

# -------------------------
# LOAD + CLEAN (robust date/value column detection)
# -------------------------
def clean_numeric(x):
    if pd.isna(x):
        return np.nan
    s = str(x).strip().replace("\xa0", "").replace(",", "")
    if s in ["", ".", "NA", "N/A", "null", "None"]:
        return np.nan
    m = re.search(r"-?\d+(\.\d+)?", s)
    if not m:
        return np.nan
    try:
        return float(m.group(0))
    except:
        return np.nan

def infer_date_col(df: pd.DataFrame) -> str:
    for c in df.columns:
        cn = c.strip().lower()
        if cn in ["date", "observation_date", "time", "timestamp"]:
            return c
        if cn == "date" or cn == "dt":
            return c
        if c.strip().upper() == "DATE":
            return c
    return df.columns[0]

def infer_value_col(df: pd.DataFrame, date_col: str, series_hint: str) -> str:
    for c in df.columns:
        if c == date_col:
            continue
        if series_hint.upper() in str(c).upper():
            return c
    for c in df.columns:
        if c == date_col:
            continue
        cn = c.strip().lower()
        if cn in ["value", "close", "last", "price", "rate"]:
            return c
    non_date = [c for c in df.columns if c != date_col]
    if len(non_date) == 1:
        return non_date[0]
    return non_date[0]

def load_series(csv_path: str, series_name: str) -> pd.Series:
    df = pd.read_csv(csv_path)
    date_col = infer_date_col(df)
    val_col  = infer_value_col(df, date_col, series_name)

    df[date_col] = pd.to_datetime(df[date_col], errors="coerce")
    df[val_col]  = df[val_col].apply(clean_numeric)

    df = df.dropna(subset=[date_col, val_col]).copy()
    df = df.sort_values(date_col).drop_duplicates(subset=[date_col], keep="last")

    s = pd.Series(df[val_col].values, index=df[date_col].values, name=series_name).astype(float)
    return s

# -------------------------
# PLOTS (black bg + watermark + final point label)
# -------------------------
def plot_black(series, title, hlines=None, final_label_fmt="{:.2f}", save_path=None):
    s = pd.Series(series).dropna()
    if s.empty:
        print(f"[plot_black] No data: {title}")
        return

    fig, ax = plt.subplots(figsize=(14,5))
    fig.patch.set_facecolor("black")
    ax.set_facecolor("black")

    ax.plot(s.index, s.values, linewidth=1.6)

    if hlines is not None:
        for yv in hlines:
            ax.axhline(yv, linewidth=1.2)

    ax.set_title(title, color="white", fontsize=14, pad=10)
    ax.tick_params(colors="white")
    for spine in ax.spines.values():
        spine.set_color("white")
    ax.grid(True, alpha=0.2)

    ax.text(0.5, 0.5, "Pearl Gray Equity and Research",
            transform=ax.transAxes, fontsize=26, color="white",
            alpha=0.08, ha="center", va="center")

    x_last = s.index[-1]
    y_last = float(s.iloc[-1])
    ax.scatter([x_last], [y_last], s=35)
    ax.annotate(final_label_fmt.format(y_last),
                xy=(x_last, y_last),
                xytext=(10, 10),
                textcoords="offset points",
                color="white",
                fontsize=11,
                bbox=dict(boxstyle="round,pad=0.25", fc="black", ec="white", alpha=0.85))

    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=200, facecolor=fig.get_facecolor(), bbox_inches="tight")
    plt.show()

def hist_kde(data, title, xlabel, bins=50, save_path=None):
    x = pd.Series(data).dropna().astype(float)
    plt.figure(figsize=(11,4))
    plt.title(title)
    plt.xlabel(xlabel)
    plt.grid(True, alpha=0.2)

    if len(x) < 2 or np.isclose(x.var(), 0.0):
        plt.text(0.5, 0.5, "Not enough variation", ha="center", va="center", transform=plt.gca().transAxes)
        plt.tight_layout()
        if save_path:
            plt.savefig(save_path, dpi=200, bbox_inches="tight")
        plt.show()
        return

    plt.hist(x.values, bins=bins, density=True, alpha=0.35)
    kde = gaussian_kde(x.values)
    grid = np.linspace(x.min(), x.max(), 400)
    plt.plot(grid, kde(grid), linewidth=2.0)
    plt.ylabel("Density")
    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=200, bbox_inches="tight")
    plt.show()

# -------------------------
# Rolling beta-hedged butterfly: dy5 ~ c + a*dy2 + b*dy10 (past window only)
# -------------------------
def rolling_beta_residual(dfd: pd.DataFrame, window: int) -> pd.DataFrame:
    idx = dfd.index
    x2  = dfd["dy2"].values
    x10 = dfd["dy10"].values
    y   = dfd["dy5"].values

    out_idx, a_list, b_list, c_list, resid_list = [], [], [], [], []

    for i in range(window, len(dfd)):
        X2  = x2[i-window:i]
        X10 = x10[i-window:i]
        Y   = y[i-window:i]

        X = np.column_stack([np.ones(window), X2, X10])
        beta = np.linalg.lstsq(X, Y, rcond=None)[0]
        c, a, b = float(beta[0]), float(beta[1]), float(beta[2])

        dy2_i, dy10_i, dy5_i = float(x2[i]), float(x10[i]), float(y[i])
        resid = dy5_i - (c + a*dy2_i + b*dy10_i)

        out_idx.append(idx[i])
        c_list.append(c)
        a_list.append(a)
        b_list.append(b)
        resid_list.append(resid)

    return pd.DataFrame(
        {"c": c_list, "a_dy2": a_list, "b_dy10": b_list, "fly_delta_bps": resid_list},
        index=pd.DatetimeIndex(out_idx)
    )

# -------------------------
# Rolling Z (shift(1), no lookahead)
# -------------------------
def compute_z(level: pd.Series, window: int) -> pd.Series:
    rm = level.shift(1).rolling(window, min_periods=window).mean()
    rs = level.shift(1).rolling(window, min_periods=window).std(ddof=1)
    z = (level - rm) / rs
    z.name = "Z"
    return z

# -------------------------
# STRATEGY: Enter at |z|>=2, Exit at z crossing 0 (full reversion)
# Single unit, both sides, non-overlapping
# Slippage: 10 bps total per completed trade (half entry day, half exit day)
# Financing: 1.77% annual, applied daily only while in position
# Daily PnL in bps: position_units * fly_delta_bps - costs
# -------------------------
def run_entry2_exit0_strategy(fly_level_bps: pd.Series,
                              fly_delta_bps: pd.Series,
                              z: pd.Series,
                              entry_level: float,
                              slippage_bps_total: float,
                              financing_annual: float,
                              trading_days: int = 252):

    df = pd.DataFrame({
        "fly_level_bps": fly_level_bps,
        "fly_delta_bps": fly_delta_bps,
        "z": z
    }).dropna()

    if df.empty:
        return {"trades": pd.DataFrame(), "daily": pd.DataFrame()}

    half_cost = slippage_bps_total / 2.0
    financing_daily_bps_per_unit = (financing_annual / trading_days) * 10000.0

    in_trade = False
    side = 0                 # +1 long fly, -1 short fly
    entry_date = None
    entry_z = None
    entry_level_value = None
    entry_cost_charged = False

    pos_units = np.zeros(len(df), dtype=int)
    pnl_bps_daily = np.zeros(len(df), dtype=float)
    cost_bps_daily = np.zeros(len(df), dtype=float)
    financing_bps_daily = np.zeros(len(df), dtype=float)

    trades = []

    for i in range(len(df)):
        zi = float(df["z"].iloc[i])
        lvl = float(df["fly_level_bps"].iloc[i])
        dte = df.index[i]

        # ENTRY (only when flat)
        if not in_trade:
            if zi >= entry_level:
                in_trade = True
                side = -1
                entry_date = dte
                entry_z = zi
                entry_level_value = lvl
                entry_cost_charged = False
            elif zi <= -entry_level:
                in_trade = True
                side = +1
                entry_date = dte
                entry_z = zi
                entry_level_value = lvl
                entry_cost_charged = False

            if in_trade and not entry_cost_charged:
                cost_bps_daily[i] += half_cost
                entry_cost_charged = True

        # EXIT when z crosses 0 (same-side)
        if in_trade:
            exit_now = False
            if side == +1 and zi >= 0.0:
                exit_now = True
            elif side == -1 and zi <= 0.0:
                exit_now = True

            if exit_now:
                cost_bps_daily[i] += half_cost

                pnl_gross = (lvl - float(entry_level_value)) * float(side)
                pnl_net = pnl_gross - float(slippage_bps_total)

                trades.append({
                    "side": "LONG" if side == +1 else "SHORT",
                    "entry_date": entry_date,
                    "exit_date": dte,
                    "days_held": int((dte - entry_date).days),
                    "entry_z": float(entry_z),
                    "exit_z": float(zi),
                    "entry_fly_level_bps": float(entry_level_value),
                    "exit_fly_level_bps": float(lvl),
                    "pnl_bps_gross": float(pnl_gross),
                    "pnl_bps_net": float(pnl_net),
                })

                in_trade = False
                side = 0
                entry_date = None
                entry_z = None
                entry_level_value = None
                entry_cost_charged = False

        current_pos = int(side) if in_trade else 0
        pos_units[i] = current_pos

        fin_cost = abs(current_pos) * financing_daily_bps_per_unit
        financing_bps_daily[i] = fin_cost

        pnl_bps_daily[i] = current_pos * float(df["fly_delta_bps"].iloc[i]) - cost_bps_daily[i] - financing_bps_daily[i]

    daily = pd.DataFrame({
        "fly_level_bps": df["fly_level_bps"],
        "fly_delta_bps": df["fly_delta_bps"],
        "z": df["z"],
        "position_units": pos_units,
        "pnl_bps": pnl_bps_daily,
        "cost_bps": cost_bps_daily,
        "financing_bps": financing_bps_daily
    }, index=df.index)

    trades = pd.DataFrame(trades)
    return {"trades": trades, "daily": daily}

# -------------------------
# Portfolio metrics
# -------------------------
def winsorize(r: pd.Series, central_mass=0.95) -> pd.Series:
    r = pd.Series(r).dropna().astype(float)
    if r.empty:
        return r
    alpha = (1.0 - central_mass) / 2.0
    lo = float(np.quantile(r, alpha))
    hi = float(np.quantile(r, 1.0 - alpha))
    return r.clip(lo, hi)

def perf_metrics(daily_ret: pd.Series, rf_annual=0.03, trading_days=252) -> dict:
    r = pd.Series(daily_ret).dropna().astype(float)
    if r.empty:
        return {"CAGR": np.nan, "MaxDD": np.nan, "Sharpe": np.nan, "Calmar": np.nan, "VaR95": np.nan}

    eq = (1.0 + r).cumprod()
    years = len(r) / trading_days
    cagr = float(eq.iloc[-1] ** (1.0/years) - 1.0) if years > 0 else np.nan

    peak = eq.cummax()
    dd = (eq / peak) - 1.0
    maxdd = float(dd.min())

    rf_daily = (1.0 + rf_annual) ** (1.0/trading_days) - 1.0
    ex = r - rf_daily
    sharpe = float(np.sqrt(trading_days) * ex.mean() / ex.std(ddof=1)) if ex.std(ddof=1) > 0 else np.nan

    calmar = float(cagr / abs(maxdd)) if maxdd < 0 else np.nan
    var95 = float(np.quantile(r, 0.05))

    return {"CAGR": cagr, "MaxDD": maxdd, "Sharpe": sharpe, "Calmar": calmar, "VaR95": var95}

def summarize_trades(trades: pd.DataFrame) -> dict:
    if trades is None or trades.empty:
        return {"N_trades": 0, "HitRate": np.nan, "Median_days": np.nan, "Mean_days": np.nan,
                "Mean_pnl_bps_net": np.nan, "Median_pnl_bps_net": np.nan}
    pnl = trades["pnl_bps_net"].astype(float)
    days = trades["days_held"].astype(float)
    return {
        "N_trades": int(len(trades)),
        "HitRate": float((pnl > 0).mean()),
        "Median_days": float(days.median()),
        "Mean_days": float(days.mean()),
        "Mean_pnl_bps_net": float(pnl.mean()),
        "Median_pnl_bps_net": float(pnl.median()),
    }

# -------------------------
# OU MLE on fly_level (dt=1 day)
# -------------------------
def ou_neg_loglik(params, X):
    kappa, mu_, sigma = params
    if kappa <= 0 or sigma <= 0:
        return 1e50
    e = np.exp(-kappa)
    v = (sigma**2 / (2.0*kappa)) * (1.0 - np.exp(-2.0*kappa))
    if not np.isfinite(v) or v <= 0:
        return 1e50
    X_t = X[:-1]
    X_tp1 = X[1:]
    m = mu_ + (X_t - mu_) * e
    ll = -0.5 * np.sum(np.log(2*np.pi*v) + ((X_tp1 - m)**2) / v)
    return -ll

def fit_ou_mle(series: pd.Series):
    X = pd.Series(series).dropna().values.astype(float)
    if len(X) < 200:
        raise ValueError("Not enough observations for OU MLE (need ~200+ daily).")
    Xt, Yt = X[:-1], X[1:]
    b = np.cov(Xt, Yt, ddof=1)[0,1] / np.var(Xt, ddof=1)
    b = float(np.clip(b, -0.999, 0.999))
    a = float(np.mean(Yt) - b*np.mean(Xt))
    k0 = max(1e-6, float(-np.log(max(b, 1e-6)))) if b > 0 else 0.05
    mu0 = float(a/(1-b)) if abs(1-b) > 1e-6 else float(np.mean(X))
    eps = Yt - (a + b*Xt)
    s2 = float(np.var(eps, ddof=1))
    denom = (1 - np.exp(-2*k0))
    sig0 = np.sqrt(max(s2 * 2*k0 / max(denom, 1e-8), 1e-8))

    x0 = np.array([k0, mu0, sig0], dtype=float)
    bounds = [(1e-10, 50.0), (None, None), (1e-10, 1e6)]
    res = minimize(ou_neg_loglik, x0=x0, args=(X,), method="L-BFGS-B", bounds=bounds)
    if not res.success:
        raise RuntimeError(f"OU MLE failed: {res.message}")
    return float(res.x[0]), float(res.x[1]), float(res.x[2])

def rolling_ou_params(series: pd.Series, window: int, step: int) -> pd.DataFrame:
    s = pd.Series(series).dropna()
    dates = s.index
    vals = s.values.astype(float)
    out_dates, out_k, out_mu, out_sig = [], [], [], []
    for start in range(0, len(vals) - window + 1, step):
        end = start + window
        sw = pd.Series(vals[start:end], index=dates[start:end])
        try:
            k, m, sig = fit_ou_mle(sw)
        except Exception:
            k, m, sig = np.nan, np.nan, np.nan
        out_dates.append(dates[end-1])
        out_k.append(k); out_mu.append(m); out_sig.append(sig)
    return pd.DataFrame({"kappa": out_k, "mu": out_mu, "sigma": out_sig}, index=pd.DatetimeIndex(out_dates))

# ============================================================
# LOAD ONCE (robust)
# ============================================================
p2  = find_file(Y2_FILE)
p5  = find_file(Y5_FILE)
p10 = find_file(Y10_FILE)

y2  = load_series(p2,  "DGS2")
y5  = load_series(p5,  "DGS5")
y10 = load_series(p10, "DGS10")

common = y2.index.intersection(y5.index).intersection(y10.index)
y2, y5, y10 = y2.loc[common], y5.loc[common], y10.loc[common]

df_all = pd.DataFrame({
    "dy2":  y2.diff()  * 100.0,
    "dy5":  y5.diff()  * 100.0,
    "dy10": y10.diff() * 100.0
}).dropna()

print("Loaded aligned daily yields:",
      df_all.index.min().date(), "->", df_all.index.max().date(),
      "| N=", len(df_all))

# ============================================================
# RUN PER REGIME
# ============================================================
master_rows = []

for regime_name, start_date, end_date in REGIMES:
    df = df_all.copy()
    if start_date is not None:
        df = df.loc[df.index >= pd.to_datetime(start_date)]
    if end_date is not None:
        df = df.loc[df.index <= pd.to_datetime(end_date)]

    print("\n==============================================")
    print("REGIME:", regime_name, "|", (start_date or "START"), "->", (end_date or "END"), "| N=", len(df))
    print("==============================================")

    if len(df) < max(ROLL_BETA_WINDOW, ROLL_Z_WINDOW) + 50:
        print("Not enough data for rolling windows. Skipping.")
        continue

    # 1) rolling hedge residual => fly_delta_bps
    roll = rolling_beta_residual(df, window=ROLL_BETA_WINDOW)
    fly_delta = roll["fly_delta_bps"].copy()
    fly_level = fly_delta.cumsum()
    fly_level.name = "fly_level_bps"

    # 2) rolling Z (no lookahead)
    z = compute_z(fly_level, window=ROLL_Z_WINDOW)

    core = pd.DataFrame({
        "fly_delta_bps": fly_delta,
        "fly_level_bps": fly_level,
        "z": z,
        "a_dy2": roll["a_dy2"],
        "b_dy10": roll["b_dy10"],
        "c": roll["c"]
    }).dropna()

    core_csv = os.path.join(OUT_DIR, f"{regime_name}_core_series.csv")
    core.to_csv(core_csv, index_label="Date")

    # 3) strategy (ENTRY 2.0, EXIT 0.0)
    res = run_entry2_exit0_strategy(
        fly_level_bps=core["fly_level_bps"],
        fly_delta_bps=core["fly_delta_bps"],
        z=core["z"],
        entry_level=float(ENTRY_LEVEL),
        slippage_bps_total=float(SLIPPAGE_BPS_PER_TRADE_TOTAL),
        financing_annual=float(FINANCING_ANNUAL),
        trading_days=252
    )
    trades = res["trades"]
    daily  = res["daily"]

    trades_csv = os.path.join(OUT_DIR, f"{regime_name}_trades_ENTRY2_EXIT0.csv")
    daily_csv  = os.path.join(OUT_DIR, f"{regime_name}_daily_ENTRY2_EXIT0.csv")
    trades.to_csv(trades_csv, index=False)
    daily.to_csv(daily_csv, index_label="Date")

    # 4) portfolio series
    daily["pnl_usd"] = (daily["pnl_bps"] / 10000.0) * NOTIONAL_USD
    daily["ret"] = daily["pnl_usd"] / NOTIONAL_USD

    met = perf_metrics(daily["ret"], rf_annual=RF_ANNUAL)
    wret = winsorize(daily["ret"], central_mass=0.95)
    met_w = perf_metrics(wret, rf_annual=RF_ANNUAL)

    tsum = summarize_trades(trades)

    # 5) density plots
    if not trades.empty:
        hist_kde(trades["days_held"], f"{regime_name} Holding Time Density (days)", "Days held",
                 save_path=os.path.join(OUT_DIR, f"{regime_name}_density_holding_days.png"))
        hist_kde(trades["pnl_bps_net"], f"{regime_name} Trade PnL Density (net bps)", "Net PnL (bps)",
                 save_path=os.path.join(OUT_DIR, f"{regime_name}_density_trade_pnl_bps.png"))
    hist_kde(core["fly_level_bps"], f"{regime_name} Fly Level Density (bps)", "Fly level (bps)",
             save_path=os.path.join(OUT_DIR, f"{regime_name}_density_fly_level.png"))

    # 6) plots: z and boundaries (entry 2, exit 0)
    plot_black(core["z"], f"{regime_name} Daily Z-score (entry {ENTRY_LEVEL} | exit {EXIT_LEVEL})",
               hlines=[-ENTRY_LEVEL, -0.75, 0.0, 0.75, ENTRY_LEVEL],
               final_label_fmt="{:.2f}",
               save_path=os.path.join(OUT_DIR, f"{regime_name}_z_black.png"))

    # optional equity curve plot
    eq = (1.0 + daily["ret"].fillna(0.0)).cumprod()
    plot_black(eq, f"{regime_name} Equity Curve (ENTRY2 EXIT0)",
               final_label_fmt="{:.4f}",
               save_path=os.path.join(OUT_DIR, f"{regime_name}_equity_black.png"))

    # 7) OU fit + half-life
    kappa, mu_ou, sigma_ou = fit_ou_mle(core["fly_level_bps"])
    half_life_days = np.log(2.0) / kappa

    # 8) rolling OU stability
    roll_ou = rolling_ou_params(core["fly_level_bps"], window=ROLL_OU_WINDOW, step=ROLL_OU_STEP)
    roll_ou_csv = os.path.join(OUT_DIR, f"{regime_name}_rolling_ou_params.csv")
    roll_ou.to_csv(roll_ou_csv, index_label="Date")

    plot_black(roll_ou["kappa"], f"{regime_name} Rolling OU kappa (window={ROLL_OU_WINDOW}d step={ROLL_OU_STEP})",
               final_label_fmt="{:.6f}",
               save_path=os.path.join(OUT_DIR, f"{regime_name}_kappa_black.png"))

    # 9) summary row
    row = {
        "Regime": regime_name,
        "Notional_USD": NOTIONAL_USD,
        "Slippage_bps_per_trade_total": SLIPPAGE_BPS_PER_TRADE_TOTAL,
        "Financing_annual": FINANCING_ANNUAL,
        "Entry_level": ENTRY_LEVEL,
        "Exit_level": EXIT_LEVEL,
        "OU_kappa": kappa,
        "OU_mu": mu_ou,
        "OU_sigma": sigma_ou,
        "OU_half_life_days": half_life_days,
        **tsum,
        **{f"PORT_{k}": v for k, v in met.items()},
        **{f"WINSOR_{k}": v for k, v in met_w.items()},
    }
    master_rows.append(row)

    print("\nSUMMARY ", regime_name)
    print(pd.DataFrame([row]).to_string(index=False))

# Save master summary
master = pd.DataFrame(master_rows)
master_csv = os.path.join(OUT_DIR, "MASTER_summary_all_regimes.csv")
master.to_csv(master_csv, index=False)

print("\nDONE.")
print("Saved outputs under:", OUT_DIR)
print("Key files:")
print(" - *_core_series.csv")
print(" - *_trades_ENTRY2_EXIT0.csv")
print(" - *_daily_ENTRY2_EXIT0.csv")
print(" - *_rolling_ou_params.csv")
print(" - MASTER_summary_all_regimes.csv")




```
