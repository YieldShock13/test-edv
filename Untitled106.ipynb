```python
# ============================================================
# JSE sectors vs All-Share: WEEKLY rolling Z-scores (sector − ALSI)
# Creates:
#   1. Main multi-line chart (all sectors)
#   2. Individual chart per sector
#   3. Watermark: "orca's capital strategies"
# Saves everything to current working directory (no /mnt/data)
# ============================================================

import os, re, glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ---- Params ----
WINDOW_WEEKS = 52
MIN_PERIODS  = 26
WATERMARK    = "orca's capital strategies"
OUT_DIR      = os.getcwd()
os.makedirs(OUT_DIR, exist_ok=True)

PRICE_CANDIDATES = ["Price", "Close", "Last", "Adj Close"]

FILES = {
    "All Share": "FTSE_JSE All Share Historical Data.csv",
    "Industrial 25": "FTSE_JSE Industrial 25 Historical Data.csv",
    "Financial 15": "FTSE_JSE Financial 15 Historical Data.csv",
    "REITs": "FTSE JSE Real Estate Investment Trusts Historical Data.csv",
    "Resource 10": "FTSE_JSE Resource 10 Historical Data.csv",
}

# ---- Helpers ----
def read_investing_csv(path):
    df = pd.read_csv(path)
    df.columns = [c.strip() for c in df.columns]
    date_col = next((c for c in df.columns if c.lower()=="date"), None)
    df[date_col] = pd.to_datetime(df[date_col], errors="coerce")
    price_col = next((c for c in PRICE_CANDIDATES if c in df.columns), None)
    if price_col is None:
        price_col = next((c for c in df.columns if c.lower() in [p.lower() for p in PRICE_CANDIDATES]), None)
    def _to_float(x):
        if isinstance(x,str):
            x = re.sub(r"[^0-9.\-]", "", x.replace(",",""))
        try: return float(x)
        except: return np.nan
    df[price_col] = df[price_col].apply(_to_float)
    s = df.dropna(subset=[date_col,price_col]).sort_values(date_col)
    s = s.set_index(date_col)[price_col].resample("W-FRI").last().dropna()
    return s.rename("price")

def weekly_log_returns(price):
    return np.log(price).diff()

def rolling_z_of_spread(sec_ret, mkt_ret, window, min_periods):
    rel = (sec_ret - mkt_ret).dropna()
    rel = rel.reindex(mkt_ret.index).dropna()
    m = rel.rolling(window, min_periods=min_periods).mean()
    s = rel.rolling(window, min_periods=min_periods).std(ddof=0)
    return (rel - m) / s

# ---- Load data ----
series = {n: read_investing_csv(v) for n,v in FILES.items()}
rets = pd.DataFrame({n: weekly_log_returns(s) for n,s in series.items()}).dropna(how="all")

alshi = rets["All Share"]
zs = {n: rolling_z_of_spread(rets[n], alshi, WINDOW_WEEKS, MIN_PERIODS)
      for n in rets.columns if n != "All Share"}
zsdf = pd.DataFrame(zs).dropna(how="all")

# ---- Plot utilities ----
def add_bands(ax):
    for lvl in [0,1,2,3,-1,-2,-3]:
        ax.axhline(y=lvl, ls="--" if lvl else "-", lw=1, alpha=0.6)

def add_watermark(ax,text=WATERMARK):
    ax.text(0.995,0.02,text,transform=ax.transAxes,
            ha="right",va="bottom",alpha=0.5,fontsize=9)

def label_latest(ax,df):
    ld = df.index.max()
    vals = df.loc[ld].dropna().sort_values()
    nudges = np.linspace(-20,20,len(vals))
    for (n,v),dy in zip(vals.items(),nudges):
        ax.annotate(f"{n}: {v:.2f}",xy=(ld,v),xytext=(10,dy),
                    textcoords="offset points",va="center",fontsize=9,
                    bbox=dict(boxstyle="round,pad=0.15",fc="white",ec="none",alpha=0.7))

# ---- MAIN CHART ----
fig, ax = plt.subplots(figsize=(14,7))
for col in zsdf.columns:
    ax.plot(zsdf.index,zsdf[col],label=col)
add_bands(ax); add_watermark(ax); label_latest(ax,zsdf)
ax.set_title(f"Rolling Z-Scores: Sector vs All-Share\nWindow={WINDOW_WEEKS}w")
ax.set_xlabel("Date"); ax.set_ylabel("Z-score (sector − All Share)")
ax.legend(ncol=2,frameon=False); ax.grid(alpha=0.3); fig.tight_layout()

main_png = os.path.join(OUT_DIR,"ZScores_AllSectors.png")
fig.savefig(main_png,dpi=150)
plt.show()

# ---- INDIVIDUAL CHARTS ----
for col in zsdf.columns:
    fig, ax = plt.subplots(figsize=(12,6))
    ax.plot(zsdf.index,zsdf[col],label=col,color="tab:blue")
    add_bands(ax); add_watermark(ax)
    if not zsdf[col].dropna().empty:
        ld = zsdf[col].dropna().index.max()
        val = zsdf[col].loc[ld]
        ax.annotate(f"{col}: {val:.2f}",xy=(ld,val),xytext=(10,0),
                    textcoords="offset points",va="center",fontsize=9,
                    bbox=dict(boxstyle="round,pad=0.15",fc="white",ec="none",alpha=0.7))
    ax.set_title(f"{col} vs All-Share (Rolling Z, {WINDOW_WEEKS}w)")
    ax.set_xlabel("Date"); ax.set_ylabel("Z-score (sector − All Share)")
    ax.grid(alpha=0.3); fig.tight_layout()
    outp = os.path.join(OUT_DIR,f"ZScore_{col.replace(' ','_')}.png")
    fig.savefig(outp,dpi=150)
    plt.show()

# ---- Save latest Z ----
latest = zsdf.loc[zsdf.index.max()].to_frame("latest_z")
csv_path = os.path.join(OUT_DIR,"Latest_ZScores.csv")
latest.to_csv(csv_path)
print(f"Saved:\n  Main: {main_png}\n  Per-sector charts: {OUT_DIR}\n  CSV: {csv_path}")





```
