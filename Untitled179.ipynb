```python
# ==========================================================
# HESTON VARIANCE DECOMPOSITION THROUGH TIME (VIX^2)
# Deterministic drift vs stochastic shock, rolling params
# SAME ANALYSIS AS BEFORE — ONLY AESTHETICS CHANGED
# Professional academic font + color scheme + styling
# ==========================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from cycler import cycler

# ---------------- PROFESSIONAL ACADEMIC LOOK (COLAB-EXISTS) ----------------
# Font: Liberation Serif is installed in Colab and looks like an academic PDF (not the ChatGPT/STIX vibe)
plt.rcParams['font.family'] = 'Liberation Serif'

# Styling: clean white background, subtle grid, non-default color cycle (NOT Matplotlib blue)
plt.rcParams['figure.dpi'] = 160
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.edgecolor'] = '#111111'
plt.rcParams['axes.linewidth'] = 0.9
plt.rcParams['axes.grid'] = True
plt.rcParams['grid.color'] = '#d0d0d0'
plt.rcParams['grid.linewidth'] = 0.6
plt.rcParams['grid.alpha'] = 0.6
plt.rcParams['xtick.color'] = '#111111'
plt.rcParams['ytick.color'] = '#111111'
plt.rcParams['text.color'] = '#111111'
plt.rcParams['axes.labelcolor'] = '#111111'
plt.rcParams['axes.titlecolor'] = '#111111'
plt.rcParams['axes.titlesize'] = 13
plt.rcParams['axes.labelsize'] = 11

# Academic palette (ColorBrewer-like, muted, readable)
plt.rcParams['axes.prop_cycle'] = cycler(color=[
    '#2C7BB6',  # muted blue
    '#D7191C',  # muted red
    '#1A9641',  # muted green
    '#FDAE61',  # muted orange
    '#6A3D9A',  # muted purple
    '#666666'   # neutral grey
])

# ---------------- LOAD DATA ----------------
df = pd.read_csv("CBOE Volatility Index Historical Data (11).csv")
df.columns = df.columns.str.strip()

df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')

# VIX^2 as variance proxy
df['VIX'] = df['Price']
df['v'] = (df['VIX'] / 100.0) ** 2
df = df[['Date', 'v']].dropna().reset_index(drop=True)

dt = 1/252  # trading-day step

# ---------------- FUNCTION: ESTIMATE HESTON-LIKE PARAMS (UNCHANGED) ----------------
def estimate_k_theta_xi(v_series, dt=1/252):
    # v_{t+1} = a + b v_t + e_t
    v_t = v_series[:-1]
    v_next = v_series[1:]

    X = sm.add_constant(v_t)
    res = sm.OLS(v_next, X).fit()

    a, b = res.params
    kappa = (1 - b) / dt

    if np.isfinite(kappa) and abs(kappa) > 1e-12:
        theta = a / (kappa * dt)
    else:
        theta = np.nan

    e = res.resid

    # diffusion scale: e_t ≈ xi * sqrt(v_t) * sqrt(dt) * z_t
    denom = np.sqrt(np.maximum(v_t, 1e-12)) * np.sqrt(dt)
    xi = np.nanstd(e / denom)

    return kappa, theta, xi

# ---------------- ROLLING ESTIMATION (UNCHANGED) ----------------
window = 252  # 1y rolling
v = df['v'].values

kappa_list, theta_list, xi_list, out_dates = [], [], [], []

for i in range(window, len(v)):
    sub = v[i-window:i]
    k, t, x = estimate_k_theta_xi(sub, dt)
    kappa_list.append(k)
    theta_list.append(t)
    xi_list.append(x)
    out_dates.append(df['Date'].iloc[i])

roll = pd.DataFrame({
    'Date': out_dates,
    'kappa': kappa_list,
    'theta': theta_list,
    'xi': xi_list
}).reset_index(drop=True)

# Align v_t and v_{t+1} to rolling dates (UNCHANGED)
roll_idx = np.arange(window, len(v))
v_t = v[roll_idx - 1]
v_next = v[roll_idx]
dv = v_next - v_t

roll['v_t'] = v_t
roll['dv'] = dv

# ---------------- DECOMPOSITION (UNCHANGED) ----------------
roll['drift'] = roll['kappa'] * (roll['theta'] - roll['v_t']) * dt
roll['shock'] = roll['dv'] - roll['drift']
roll['shock_std'] = roll['shock'] / (roll['xi'] * np.sqrt(np.maximum(roll['v_t'], 1e-12)) * np.sqrt(dt))

# ---------------- STATS (UNCHANGED: real summary table) ----------------
def summarize(series, name):
    s = series.dropna()
    return {
        "name": name,
        "count": int(s.shape[0]),
        "mean": float(s.mean()),
        "std": float(s.std(ddof=1)),
        "p01": float(s.quantile(0.01)),
        "p05": float(s.quantile(0.05)),
        "p50": float(s.quantile(0.50)),
        "p95": float(s.quantile(0.95)),
        "p99": float(s.quantile(0.99)),
    }

stats = pd.DataFrame([
    summarize(roll["kappa"], "kappa"),
    summarize(roll["theta"], "theta"),
    summarize(roll["xi"], "xi"),
    summarize(roll["drift"], "drift (expected dv)"),
    summarize(roll["shock"], "shock (dv - drift)"),
    summarize(roll["shock_std"], "standardized shock z"),
])

print("\n===== ROLLING PARAMETER + DECOMPOSITION STATS =====")
print(stats.to_string(index=False))

# ---------------- PLOT-ONLY WINSORIZATION (UNCHANGED INTENT) ----------------
def winsorize(s, q=0.005):
    lo, hi = s.quantile(q), s.quantile(1-q)
    return s.clip(lo, hi)

roll_plot = roll.copy()
roll_plot["kappa_w"] = winsorize(roll_plot["kappa"], 0.01)
roll_plot["theta_w"] = winsorize(roll_plot["theta"], 0.01)
roll_plot["xi_w"] = winsorize(roll_plot["xi"], 0.01)
roll_plot["drift_w"] = winsorize(roll_plot["drift"], 0.01)
roll_plot["shock_w"] = winsorize(roll_plot["shock"], 0.01)
roll_plot["shock_std_w"] = winsorize(roll_plot["shock_std"], 0.01)

# ---------------- Helper: label last point ----------------
def label_last(ax, x, y, fmt="{:.4f}", dx=6, dy=6):
    ax.scatter([x[-1]], [y[-1]], zorder=5, s=18, color='#111111')
    ax.annotate(fmt.format(y[-1]),
                (x[-1], y[-1]),
                textcoords="offset points",
                xytext=(dx, dy),
                ha="left", va="bottom")

def finish_axes(ax):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

# ==========================================================
# CHART A: v_t (variance) with last label
# ==========================================================
fig, ax = plt.subplots(figsize=(10.8, 3.8))
ax.plot(df["Date"], df["v"], linewidth=1.1)
ax.set_title("VIX Implied Variance  (v = (VIX/100)²)")
ax.set_xlabel("Date"); ax.set_ylabel("Variance v")
label_last(ax, df["Date"].values, df["v"].values, fmt="{:.4f}")
finish_axes(ax)
plt.tight_layout()
plt.show()

# ==========================================================
# CHART B: Rolling kappa, theta, xi (winsorized for readability)
# ==========================================================
fig, ax = plt.subplots(figsize=(10.8, 3.8))
ax.plot(roll_plot["Date"], roll_plot["kappa_w"], linewidth=1.1)
ax.set_title("Mean Reversion Speed (κ) — Rolling 1Y")
ax.set_xlabel("Date"); ax.set_ylabel("κ")
label_last(ax, roll_plot["Date"].values, roll_plot["kappa_w"].values, fmt="{:.2f}")
finish_axes(ax)
plt.tight_layout()
plt.show()

fig, ax = plt.subplots(figsize=(10.8, 3.8))
ax.plot(roll_plot["Date"], roll_plot["theta_w"], linewidth=1.1)
ax.set_title("Long-Run Variance (θ) — Rolling 1Y")
ax.set_xlabel("Date"); ax.set_ylabel("θ")
label_last(ax, roll_plot["Date"].values, roll_plot["theta_w"].values, fmt="{:.4f}")
finish_axes(ax)
plt.tight_layout()
plt.show()

fig, ax = plt.subplots(figsize=(10.8, 3.8))
ax.plot(roll_plot["Date"], roll_plot["xi_w"], linewidth=1.1)
ax.set_title("Vol-of-Vol (ξ) — Rolling 1Y")
ax.set_xlabel("Date"); ax.set_ylabel("ξ")
label_last(ax, roll_plot["Date"].values, roll_plot["xi_w"].values, fmt="{:.3f}")
finish_axes(ax)
plt.tight_layout()
plt.show()

# ==========================================================
# CHART C: Drift vs Shock THROUGH TIME (winsorized for readability)
# Use line styles to distinguish deterministically (academic style)
# ==========================================================
fig, ax = plt.subplots(figsize=(10.8, 3.8))
ax.plot(roll_plot["Date"], roll_plot["drift_w"], linewidth=1.1, linestyle='-', label="Deterministic drift (expected Δv)")
ax.plot(roll_plot["Date"], roll_plot["shock_w"], linewidth=1.0, linestyle='--', label="Stochastic shock (Δv − drift)")
ax.set_title("Variance Change Decomposition — Rolling 1Y Parameters")
ax.set_xlabel("Date"); ax.set_ylabel("Component value")
ax.legend(frameon=False, loc='upper right')
label_last(ax, roll_plot["Date"].values, roll_plot["drift_w"].values, fmt="{:.5f}", dx=6, dy=6)
label_last(ax, roll_plot["Date"].values, roll_plot["shock_w"].values, fmt="{:.5f}", dx=6, dy=-14)
finish_axes(ax)
plt.tight_layout()
plt.show()

# ==========================================================
# CHART D: Standardized shocks z_t (winsorized)
# ==========================================================
fig, ax = plt.subplots(figsize=(10.8, 3.8))
ax.plot(roll_plot["Date"], roll_plot["shock_std_w"], linewidth=1.0, color='#333333')
ax.axhline(0, linewidth=0.9, color='#111111')
ax.set_title("Standardized Shock  zₜ = shock / (ξ √v √dt)")
ax.set_xlabel("Date"); ax.set_ylabel("zₜ")
label_last(ax, roll_plot["Date"].values, roll_plot["shock_std_w"].values, fmt="{:.2f}")
finish_axes(ax)
plt.tight_layout()
plt.show()
```
