```python
# === 1. Install Required Packages ===
!pip install fredapi keras --quiet

# === 2. Imports ===
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from keras.models import Sequential
from keras.layers import LSTM, Dense
from fredapi import Fred
import datetime

# === 3. Setup FRED API ===
fred = Fred(api_key='eb09beeb04065b310e4dc47320ab0fce')

# === 4. Pull NASDAQ Daily Data (Last 10Y) ===
end = datetime.datetime.today()
start = end - datetime.timedelta(days=365 * 10)
df = fred.get_series('NASDAQCOM', observation_start=start, observation_end=end).to_frame(name='Price')
df.index = pd.to_datetime(df.index)
df = df.asfreq('D').ffill()  # Fill weekends/holidays with prior value

# === 5. Preprocessing ===
scaler = MinMaxScaler()
scaled = scaler.fit_transform(df[['Price']])

# === 6. Create Sequences: 60-day input → predict 30 days ahead ===
def create_sequences(data, input_len=60, forecast_horizon=30):
    X, y = [], []
    for i in range(input_len, len(data) - forecast_horizon):
        X.append(data[i - input_len:i])
        y.append(data[i + forecast_horizon])
    return np.array(X), np.array(y)

X, y = create_sequences(scaled, input_len=60, forecast_horizon=30)
X = X.reshape((X.shape[0], X.shape[1], 1))

# === 7. Build & Train LSTM ===
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(60, 1)))
model.add(LSTM(units=50))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
model.fit(X, y, epochs=10, batch_size=32, verbose=1)

# === 8. Predict & Inverse Transform ===
predicted = model.predict(X)
predicted_rescaled = scaler.inverse_transform(predicted)
y_rescaled = scaler.inverse_transform(y)

# === 9. Evaluation (in % terms) ===
mse_raw = mean_squared_error(y_rescaled, predicted_rescaled)
rmse_raw = np.sqrt(mse_raw)
r2 = r2_score(y_rescaled, predicted_rescaled)
avg_price = np.mean(y_rescaled)

rmse_pct = (rmse_raw / avg_price) * 100
mse_pct = (mse_raw / (avg_price**2)) * 100
final_day = df.index[-1].strftime("%Y-%m-%d")

# === 10. Plot Forecast ===
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(14, 6))
ax.plot(df.index[90:90+len(y_rescaled)], y_rescaled, label='Actual (t+30)', color='white')
ax.plot(df.index[90:90+len(predicted_rescaled)], predicted_rescaled, label='Predicted (t+30)', color='gray', alpha=0.8)
ax.set_title("NASDAQ Forecast (30-Day Ahead) — LSTM Model", fontsize=14, weight='bold')
ax.set_xlabel("Date")
ax.set_ylabel("Index Level")
ax.legend(loc='upper left')
ax.grid(True, color='gray', linestyle='--', linewidth=0.3)
fig.text(0.5, 0.02, 'Pearl Gray Equity and Research', ha='center', fontsize=10, color='white', alpha=0.9)
plt.tight_layout(rect=[0, 0.04, 1, 0.96])
plt.show()

# === 11. Summary Table (in %) with Final Date Used ===
import seaborn as sns
summary_df = pd.DataFrame({
    'Metric': ['RMSE (%)', 'MSE (%)', 'R²', 'Time Period', 'Final Date Used'],
    'Value': [
        f'{rmse_pct:.2f}%',
        f'{mse_pct:.2f}%',
        f'{r2:.4f}',
        f'{df.index[0].strftime("%Y-%m")} to {df.index[-1].strftime("%Y-%m")}',
        final_day
    ]
})

fig, ax = plt.subplots(figsize=(8, 3))
plt.axis('off')
table = plt.table(cellText=summary_df.values,
                  colLabels=summary_df.columns,
                  cellLoc='center',
                  loc='center',
                  colColours=["#333333", "#333333"],
                  cellColours=[["#1a1a1a"]*2]*len(summary_df),
                  colLoc='center')
table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1.2, 1.5)
plt.title("LSTM Forecast — Performance Summary", fontsize=14, weight='bold')
plt.figtext(0.5, 0.97, 'Pearl Gray Equity and Research', ha='center', fontsize=10, color='white', alpha=0.85)
plt.tight_layout()
plt.show()

















```
