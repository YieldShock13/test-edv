```python
# === Install Required Libraries ===
!pip install fredapi yfinance tensorflow --quiet

# === Import Libraries ===
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from keras.models import Sequential
from keras.layers import LSTM, Dense
from fredapi import Fred

# === Load NASDAQ from FRED ===
fred = Fred(api_key='eb09beeb04065b310e4dc47320ab0fce')
nasdaq = fred.get_series('NASDAQCOM', observation_start='2015-01-01', observation_end='2025-07-25')
nasdaq = nasdaq.to_frame(name='NASDAQ')
nasdaq.index.name = 'Date'

# === Create Fake Macro Factors (replace with real if needed) ===
np.random.seed(42)
macros = pd.DataFrame({
    'FedFunds': np.random.rand(len(nasdaq)) * 5,
    '10Y_Yield': np.random.rand(len(nasdaq)) * 5 + 1,
    'Unemployment': np.random.rand(len(nasdaq)) * 3 + 4,
    'LEI': np.random.rand(len(nasdaq)) * 100,
    'Sentiment': np.random.rand(len(nasdaq)) * 100
}, index=nasdaq.index)

data = pd.concat([nasdaq, macros], axis=1).dropna()

# === Normalize ===
scaler = MinMaxScaler()
scaled = pd.DataFrame(scaler.fit_transform(data), columns=data.columns, index=data.index)

# === Sequence Builder ===
def create_sequences(data, target_col, input_len=60, forecast_horizon=30):
    X, y = [], []
    for i in range(input_len, len(data) - forecast_horizon):
        X.append(data.iloc[i - input_len:i].values)
        y.append(data.iloc[i + forecast_horizon][target_col])
    return np.array(X), np.array(y)

X, y = create_sequences(scaled, 'NASDAQ', 60, 30)

# === LSTM Model ===
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(X.shape[1], X.shape[2])))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
model.fit(X, y, epochs=5, batch_size=32, verbose=0)

# === Predict and Inverse Transform ===
pred = model.predict(X)
pred_rescaled = scaler.inverse_transform(
    np.hstack((pred, np.zeros((pred.shape[0], len(data.columns) - 1))))
)[:, 0]
y_true = scaler.inverse_transform(
    np.hstack((y.reshape(-1, 1), np.zeros((y.shape[0], len(data.columns) - 1))))
)[:, 0]

# === Metrics ===
rmse = np.sqrt(mean_squared_error(y_true, pred_rescaled))
mse = mean_squared_error(y_true, pred_rescaled)
r2 = r2_score(y_true, pred_rescaled)
n = len(y_true)
k = X.shape[1] * X.shape[2]
adj_r2 = 1 - ((1 - r2) * (n - 1)) / (n - k - 1)

# === Convert to Percentage ===
mean_level = np.mean(y_true)
rmse_pct = (rmse / mean_level) * 100
mse_pct = (mse / (mean_level ** 2)) * 100
r2_pct = r2 * 100
adj_r2_pct = adj_r2 * 100

# === Plot with Style ===
fig, ax = plt.subplots(figsize=(14, 6))
fig.patch.set_facecolor('black')
ax.set_facecolor('black')
ax.plot(data.index[-len(y_true):], y_true, label='Actual NASDAQ', color='white', linewidth=2)
ax.plot(data.index[-len(y_true):], pred_rescaled, label='Predicted NASDAQ (In-Sample)', color='lime', linewidth=2)
ax.spines['bottom'].set_color('white')
ax.spines['top'].set_color('black')
ax.spines['right'].set_color('black')
ax.spines['left'].set_color('white')
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')
ax.set_title("NASDAQ Index Prediction: In-Sample Forecast (2015–2025)", fontsize=16, fontweight='bold', color='white')
ax.set_xlabel("Date", fontsize=12, fontweight='bold', color='white')
ax.set_ylabel("Index Level", fontsize=12, fontweight='bold', color='white')
fig.text(0.01, 0.97, 'Pearl Gray Equity and Research', fontsize=10, fontweight='bold', color='white')
ax.legend(loc="upper left", facecolor='black', edgecolor='white', framealpha=1, fontsize=10)
plt.tight_layout()
plt.show()

# === Summary Table ===
summary = pd.DataFrame({
    "Metric": ["RMSE (%)", "MSE (%)", "R² (%)", "Adjusted R² (%)"],
    "Value": [f"{rmse_pct:.2f}", f"{mse_pct:.2f}", f"{r2_pct:.2f}", f"{adj_r2_pct:.2f}"]
})
print("\nSummary Results (% Terms):")
print(summary.to_string(index=False))




```
