```python
# === Install Required Libraries ===
!pip install fredapi tensorflow --quiet

# === Imports ===
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from keras.models import Sequential
from keras.layers import LSTM, Dense, Input
from fredapi import Fred

# === Load FRED Data ===
fred = Fred(api_key='eb09beeb04065b310e4dc47320ab0fce')
start_date = '2015-01-01'
end_date = '2025-07-24'

nasdaq = fred.get_series('NASDAQCOM', observation_start=start_date, observation_end=end_date).to_frame(name='NASDAQ')
fedfunds = fred.get_series('FEDFUNDS', observation_start=start_date, observation_end=end_date).to_frame(name='FedFunds')
yield_10y = fred.get_series('GS10', observation_start=start_date, observation_end=end_date).to_frame(name='10Y_Yield')
lei = fred.get_series('USSLIND', observation_start=start_date, observation_end=end_date).to_frame(name='LEI')
gdp_yoy = fred.get_series('A191RL1Q225SBEA', observation_start=start_date, observation_end=end_date).to_frame(name='GDP_YoY')

nasdaq.index.name = 'Date'
fedfunds = fedfunds.reindex(nasdaq.index).ffill()
yield_10y = yield_10y.reindex(nasdaq.index).ffill()
lei = lei.reindex(nasdaq.index).ffill()
gdp_yoy.index = gdp_yoy.index.to_period('Q').to_timestamp()
gdp_yoy = gdp_yoy.reindex(nasdaq.index).ffill()

data = pd.concat([nasdaq, fedfunds, yield_10y, lei, gdp_yoy], axis=1).dropna()

# === Normalize ===
scaler = MinMaxScaler()
scaled = pd.DataFrame(scaler.fit_transform(data), columns=data.columns, index=data.index)

# === Create Sequences ===
lookback = 60
horizon = 30
X, y = [], []

for i in range(lookback, len(scaled) - horizon):
    X.append(scaled.iloc[i - lookback:i].values)
    y.append(scaled.iloc[i:i + horizon]['NASDAQ'].values)

X = np.array(X)
y = np.array(y)

# === Build and Train LSTM ===
model = Sequential()
model.add(Input(shape=(lookback, X.shape[2])))
model.add(LSTM(50, return_sequences=True))
model.add(LSTM(50))
model.add(Dense(horizon))
model.compile(optimizer='adam', loss='mse')
model.fit(X, y, epochs=5, batch_size=32, verbose=0)

# === Inverse Transform Helper ===
def inverse_sequence(scaled_seq):
    pad = np.zeros((len(scaled_seq), scaled.shape[1]))
    pad[:, 0] = scaled_seq
    return scaler.inverse_transform(pad)[:, 0]

# === In-Sample Predictions ===
y_pred_scaled = model.predict(X, verbose=0)
y_pred = np.array([inverse_sequence(seq) for seq in y_pred_scaled])
y_true = np.array([inverse_sequence(seq) for seq in y])

# === Flatten for Metrics ===
y_pred_flat = y_pred.flatten()
y_true_flat = y_true.flatten()

# === Metrics (in %) ===
r2 = r2_score(y_true_flat, y_pred_flat)
n = len(y_true_flat)
k = X.shape[2]
adj_r2 = 1 - (1 - r2) * (n - 1) / (n - k - 1)
rmse = np.sqrt(mean_squared_error(y_true_flat, y_pred_flat))
mse = mean_squared_error(y_true_flat, y_pred_flat)
nasdaq_mean = np.mean(y_true_flat)
rmse_pct = (rmse / nasdaq_mean) * 100
mse_pct = (mse / (nasdaq_mean ** 2)) * 100

# === Styled Metrics Table ===
metrics_df = pd.DataFrame({
    "Metric": ["RMSE (%)", "MSE (%)", "R²", "Adjusted R²"],
    "Value": [f"{rmse_pct:.2f}%", f"{mse_pct:.2f}%", f"{r2:.4f}", f"{adj_r2:.4f}"]
})

fig, ax = plt.subplots(figsize=(6, 2))
fig.patch.set_facecolor('black')
ax.axis('off')
table = ax.table(cellText=metrics_df.values, colLabels=metrics_df.columns, loc='center', cellLoc='center')
table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1.2, 1.5)

for (i, j), cell in table.get_celld().items():
    cell.set_edgecolor('black')
    if i == 0:
        cell.set_facecolor('#333333')
        cell.set_text_props(color='white', weight='bold')
    else:
        cell.set_facecolor('#1a1a1a')
        cell.set_text_props(color='white')

plt.tight_layout()
plt.show()




















```
