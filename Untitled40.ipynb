```python
# === Install Required Libraries ===
!pip install fredapi tensorflow --quiet

# === Import Libraries ===
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from keras.models import Sequential
from keras.layers import LSTM, Dense, Input
from fredapi import Fred

# === Load Data from FRED ===
fred = Fred(api_key='eb09beeb04065b310e4dc47320ab0fce')
start_date = '2015-01-01'
end_date = '2025-07-24'  # changed from 2025-07-25 to 2025-07-24

# === Pull FRED Series ===
nasdaq = fred.get_series('NASDAQCOM', observation_start=start_date, observation_end=end_date).to_frame(name='NASDAQ')
fedfunds = fred.get_series('FEDFUNDS', observation_start=start_date, observation_end=end_date).to_frame(name='FedFunds')
yield_10y = fred.get_series('GS10', observation_start=start_date, observation_end=end_date).to_frame(name='10Y_Yield')
lei = fred.get_series('USSLIND', observation_start=start_date, observation_end=end_date).to_frame(name='LEI')
gdp_yoy = fred.get_series('A191RL1Q225SBEA', observation_start=start_date, observation_end=end_date).to_frame(name='GDP_YoY')

# === Align and Forward-fill to Daily ===
nasdaq.index.name = 'Date'
fedfunds = fedfunds.reindex(nasdaq.index).ffill()
yield_10y = yield_10y.reindex(nasdaq.index).ffill()
lei = lei.reindex(nasdaq.index).ffill()
gdp_yoy.index = gdp_yoy.index.to_period('Q').to_timestamp()
gdp_yoy = gdp_yoy.reindex(nasdaq.index).ffill()

# === Merge ===
data = pd.concat([nasdaq, fedfunds, yield_10y, lei, gdp_yoy], axis=1).dropna()

# === Normalize ===
scaler = MinMaxScaler()
scaled = pd.DataFrame(scaler.fit_transform(data), columns=data.columns, index=data.index)

# === Prepare Sequences ===
def create_sequences(data, target_col, input_len=60):
    X, y = [], []
    for i in range(input_len, len(data) - 1):
        X.append(data.iloc[i - input_len:i].values)
        y.append(data.iloc[i + 1][target_col])
    return np.array(X), np.array(y)

input_len = 60
X_train, y_train = create_sequences(scaled, 'NASDAQ', input_len)
last_block = scaled.iloc[-input_len:].values.reshape(1, input_len, scaled.shape[1])

# === LSTM Model ===
model = Sequential()
model.add(Input(shape=(X_train.shape[1], X_train.shape[2])))
model.add(LSTM(50, return_sequences=True))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
model.fit(X_train, y_train, epochs=5, batch_size=32, verbose=0)

# === Forecast Next 30 Days (starting at 2025-07-24 end)
forecast_scaled = []
current_input = last_block.copy()
for _ in range(30):
    next_pred = model.predict(current_input, verbose=0)
    forecast_scaled.append(next_pred[0, 0])
    next_row = np.hstack((next_pred.flatten(), current_input[0, -1, 1:]))
    current_input = np.append(current_input[:, 1:, :], next_row.reshape(1, 1, -1), axis=1)

# === Inverse Transform Forecast ===
pad = np.zeros((30, len(data.columns)))
pad[:, 0] = forecast_scaled
forecast_rescaled = scaler.inverse_transform(pad)[:, 0]
forecast_dates = pd.date_range(start='2025-07-24', periods=30, freq='D')

# === Evaluate Performance ===
y_pred_train = model.predict(X_train, verbose=0)
y_train_inv = scaler.inverse_transform(np.hstack([y_train.reshape(-1, 1)] + [np.zeros((len(y_train), data.shape[1]-1))]))[:, 0]
y_pred_inv = scaler.inverse_transform(np.hstack([y_pred_train] + [np.zeros((len(y_pred_train), data.shape[1]-1))]))[:, 0]
rmse = np.sqrt(mean_squared_error(y_train_inv, y_pred_inv))
mse = mean_squared_error(y_train_inv, y_pred_inv)
r2 = r2_score(y_train_inv, y_pred_inv)

# === Plot Forecast ===
fig, ax = plt.subplots(figsize=(14, 6))
fig.patch.set_facecolor('black')
ax.set_facecolor('black')

ax.plot(data.index, data['NASDAQ'], color='gray', label='Training Data', linewidth=1)
ax.plot(data.index[-60:], data['NASDAQ'].iloc[-60:], color='white', linewidth=2, label='Lookback Input (60d)')
ax.plot(forecast_dates, forecast_rescaled, color='lime', linewidth=2, label='Forecast (Next 30 Days)')

# Label only the final forecast value
ax.text(forecast_dates[-1], forecast_rescaled[-1], f"{forecast_rescaled[-1]:,.3f}",
        color='lime', fontsize=10, ha='left', va='bottom', fontweight='bold')

# Labels and styling
ax.set_title("NASDAQ LSTM Forecast  August 2025 Projection", fontsize=16, fontweight='bold', color='white')
ax.set_xlabel("Date", fontsize=12, fontweight='bold', color='white')
ax.set_ylabel("Index Level", fontsize=12, fontweight='bold', color='white')
ax.spines['bottom'].set_color('white')
ax.spines['left'].set_color('white')
ax.tick_params(colors='white')
ax.legend(facecolor='black', edgecolor='white', fontsize=10, loc='upper left', labelcolor='white')
fig.text(0.01, 0.97, 'Pearl Gray Equity and Research', fontsize=10, fontweight='bold', color='white')
plt.tight_layout()
plt.show()

# === Performance Table ===
metrics = {
    "RMSE (%)": f"{(rmse / y_train_inv.mean()) * 100:.2f}%",
    "MSE (%)": f"{(mse / y_train_inv.mean()**2) * 100:.2f}%",
    "R": f"{r2:.4f}",
    "Time Period": f"{data.index[0].strftime('%Y-%m')} to {data.index[-1].strftime('%Y-%m')}",
    "Final Date Used": data.index[-1].strftime('%Y-%m-%d')
}
df_metrics = pd.DataFrame(list(metrics.items()), columns=["Metric", "Value"])

# === Styled Table ===
fig, ax = plt.subplots(figsize=(8, 2))
fig.patch.set_facecolor('black')
ax.axis('off')
table = ax.table(cellText=df_metrics.values, colLabels=df_metrics.columns, loc='center', cellLoc='center')
table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1.2, 1.8)
for (i, j), cell in table.get_celld().items():
    cell.set_edgecolor('black')
    cell.set_linewidth(1.2)
    if i == 0:
        cell.set_facecolor('#333333')
        cell.set_text_props(weight='bold', color='white')
    else:
        cell.set_facecolor('#1a1a1a')
        cell.set_text_props(color='white')

plt.title("Pearl Gray Equity and Research\nLSTM Forecast  Performance Summary", fontsize=14, fontweight='bold', color='white', pad=20)
plt.tight_layout()
plt.show()









```
