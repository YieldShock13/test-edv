```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
    "font.family": "serif",
    "figure.facecolor": "white",
    "axes.facecolor": "white"
})
WATERMARK = "Orca's Capital Strategies"

def load_fred_csv(fname, col):
    df = pd.read_csv(fname)
    df["observation_date"] = pd.to_datetime(df["observation_date"])
    df[col] = pd.to_numeric(df[col], errors="coerce")
    return df.set_index("observation_date")[[col]].sort_index()

ffr   = load_fred_csv("FEDFUNDS (6).csv", "FEDFUNDS")
un    = load_fred_csv("UNRATE.csv", "UNRATE")
rs    = load_fred_csv("RSXFS.csv", "RSXFS")
pcepi = load_fred_csv("PCEPI.csv", "PCEPI")
gdp   = load_fred_csv("GDP.csv", "GDP")

def q_last(df):
    out = df.resample("QE").last()
    out.index = out.index.to_period("Q")
    return out

ffr_q   = q_last(ffr)
un_q    = q_last(un)
rs_q    = q_last(rs)
pcepi_q = q_last(pcepi)

gdp_q = gdp.copy()
gdp_q.index = gdp_q.index.to_period("Q")

q = pd.concat([ffr_q, un_q, rs_q, pcepi_q, gdp_q], axis=1).dropna()
q.index = q.index.to_timestamp("Q")

q["GDP_g_bp"] = 10000 * np.log(q["GDP"]).diff()                         # bp q/q
q["RS_g_bp"]  = 10000 * np.log(q["RSXFS"]).diff()                       # bp q/q
q["INF_yoy_bp"] = 10000 * (np.log(q["PCEPI"]) - np.log(q["PCEPI"]).shift(4))  # bp y/y

q = q.dropna()

ROLL_Q = 8
q["ffr_peak"] = q["FEDFUNDS"].rolling(ROLL_Q, min_periods=1).max()
q["cum_easing"] = q["FEDFUNDS"] - q["ffr_peak"]

STATE_LEVELS = [-1.0, -2.0, -3.0]   # 100/200/300 bps
MAX_H = 12

episodes = []
for lvl in STATE_LEVELS:
    mask = q["cum_easing"] <= lvl
    for T in q.index[mask]:
        base_un  = q.loc[T, "UNRATE"]
        base_inf = q.loc[T, "INF_yoy_bp"]
        for h in range(MAX_H + 1):
            t = T + pd.offsets.QuarterEnd(h)
            if t not in q.index:
                continue
            episodes.append({
                "state": f"{int(abs(lvl)*100)}bps",
                "T": T,
                "h": h,
                "dUN_pp": q.loc[t, "UNRATE"] - base_un,          # percentage points
                "RS_g_bp": q.loc[t, "RS_g_bp"],                 # bp q/q
                "GDP_g_bp": q.loc[t, "GDP_g_bp"],               # bp q/q
                "dINF_bp": q.loc[t, "INF_yoy_bp"] - base_inf    # bp change in y/y
            })

evt = pd.DataFrame(episodes)
if evt.empty:
    raise ValueError("No quarterly easing states found  adjust ROLL_Q, thresholds, or check data coverage.")

def plot_state(df, y, title, ylabel):
    fig, ax = plt.subplots(figsize=(10, 5))
    for T, g in df.groupby("T"):
        ax.plot(g["h"], g[y], color="gray", alpha=0.2, linewidth=1)
    avg = df.groupby("h")[y].mean()
    ax.plot(avg.index, avg.values, color="black", linewidth=3, label="Average")
    ax.text(avg.index[-1], avg.values[-1], f"{avg.values[-1]:.2f}", ha="left", va="center", fontsize=9)
    ax.axhline(0, linestyle="--", color="black", linewidth=1)
    ax.set_title(title)
    ax.set_xlabel("Quarters after state holds")
    ax.set_ylabel(ylabel)
    ax.legend(frameon=False)
    fig.text(0.5, 0.02, WATERMARK, ha="center", fontsize=9)
    plt.tight_layout()
    plt.show()

for s in sorted(evt["state"].unique(), key=lambda x: int(x.replace("bps", ""))):
    d = evt[evt["state"] == s]
    plot_state(d, "dUN_pp",   f"Unemployment once easing  {s}", "ppt change from baseline quarter")
    plot_state(d, "RS_g_bp",  f"Retail sales growth once easing  {s}", "bp q/q (log change)")
    plot_state(d, "GDP_g_bp", f"GDP growth once easing  {s}", "bp q/q (log change)")
    plot_state(d, "dINF_bp",  f"Inflation once easing  {s}", "bp change in y/y (PCEPI)")

fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(q.index, q["FEDFUNDS"], color="black", linewidth=2)

ymin, ymax = q["FEDFUNDS"].min(), q["FEDFUNDS"].max()
ax.fill_between(q.index, ymin, ymax, where=q["cum_easing"] <= -1.0, alpha=0.15, color="grey", label="100 bp")
ax.fill_between(q.index, ymin, ymax, where=q["cum_easing"] <= -2.0, alpha=0.30, color="grey", label="200 bp")
ax.fill_between(q.index, ymin, ymax, where=q["cum_easing"] <= -3.0, alpha=0.45, color="grey", label="300 bp")

ax.set_title("Fed Funds Rate with Easing States (rolling-peak definition)")
ax.set_ylabel("Fed Funds (%)")
ax.legend(frameon=False)
fig.text(0.5, 0.02, WATERMARK, ha="center", fontsize=9)
plt.tight_layout()
plt.show()








```
