```python
# ==========================================================
# ADR vs Local StatArb — IMPUY (ADR, USD) vs IMPJ (Local)
# Daily Open→Close, β-hedged, residual-z signal (no explicit FX)
# Splits: strict detect + back-adjust; blackout T & T+1
# Thresholds: 0.00, 0.25, 0.50, 1.00   |   Stops: 2.5%, 5.0%, 7.5%
# Costs: round-trip same day across 2 legs ≈ 4 × bps/side
# Watermark: "Orca's Capital Strategies"
# ==========================================================
import os, glob, math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import statsmodels.api as sm
from statsmodels.regression.rolling import RollingOLS

# ---------------- Config ----------------
WATERMARK = "Orca's Capital Strategies"
DATE_FMT_D = DateFormatter("%Y-%m-%d")

FN_ADR    = "IMPUY Historical Data.csv"      # ADR (USD)
FN_LOCAL  = "IMPJ Historical Data.csv"       # Local listing

ROLL_WINDOW = 60              # days for rolling α,β,σ
Z_THRESHOLDS = [0.00, 0.25, 0.50, 1.00]
STOP_LIST = [0.025, 0.050, 0.075]  # daily stop-loss caps (2.5%, 5%, 7.5%), no upside cap

HEDGE_MODE = "beta"           # "beta" for β-hedged notional, "equal" for equal-notional
TC_BPS_PER_SIDE = 3           # per-side transaction cost (bps). Active day ≈ 4 * TC bps

# Splits detection — conservative
ADJUST_SPLITS = True
SPLIT_FACTORS = [2, 3, 4, 5, 10]
SPLIT_TOL = 0.02              # ±2% tolerance around integer ratios (both Open & Close)
SPLIT_MIN_JUMP = 0.45         # require ≥45% jump day-over-day
SPLIT_MIN_GAP_DAYS = 60       # minimum gap between valid split detections on same leg
SPLIT_PERSIST_DAYS = 2        # require post-event levels to stay consistent ≥2 days

# ---------------- Helpers ----------------
def resolve_path(filename):
    for p in (f"/content/{filename}", f"/mnt/data/{filename}", filename):
        if os.path.isfile(p): return p
    for base in ("/content", "/mnt/data", "."):
        hits = glob.glob(os.path.join(base, "**", filename), recursive=True)
        if hits: return hits[0]
    raise FileNotFoundError(f"Could not find: {filename}")

def robust_date(s: pd.Series) -> pd.Series:
    s = s.astype(str)
    d1 = pd.to_datetime(s, dayfirst=True, errors="coerce")
    d2 = pd.to_datetime(s, dayfirst=False, errors="coerce")
    return d1 if d1.max() >= d2.max() else d2

def parse_number(series: pd.Series) -> pd.Series:
    s = series.astype(str).str.strip().replace({"-": np.nan})
    s = s.str.replace("%","",regex=False)
    # detect decimal style
    smpl = s.sample(min(200, len(s)), random_state=0)
    comma_no_dot = ((smpl.str.contains(",")) & ~smpl.str.contains(r"\.")).mean()
    if comma_no_dot > 0.6:
        s = s.str.replace(".", "", regex=False)
        s = s.str.replace(",", ".", regex=False)
    else:
        s = s.str.replace(",", "", regex=False)
    return pd.to_numeric(s, errors="coerce")

def load_daily_ohlc(path: str, prefix: str) -> pd.DataFrame:
    df = pd.read_csv(path)
    df.columns = [c.strip() for c in df.columns]
    # many "Historical Data" files have "Price"(close) and "Open"
    if not {"Date","Open"}.issubset(df.columns) or ("Price" not in df.columns and "Close" not in df.columns):
        raise ValueError(f"{path} must have Date, Open and Price/Close columns. Got: {df.columns.tolist()}")
    df["Date"]  = robust_date(df["Date"])
    df["Open"]  = parse_number(df["Open"])
    close_col = "Price" if "Price" in df.columns else "Close"
    df["Close"] = parse_number(df[close_col])
    out = df.dropna(subset=["Date","Open","Close"]).sort_values("Date").set_index("Date")
    out = out[~out.index.duplicated(keep="last")]
    # keep only the two columns to avoid join overlap
    return out.rename(columns={"Open": f"{prefix}_open", "Close": f"{prefix}_close"})[[f"{prefix}_open", f"{prefix}_close"]]

def _near(x, target, tol): return abs(x - target) <= tol

def detect_splits_strict(df: pd.DataFrame, prefix: str):
    """
    Detect splits/reverse-splits using both Open and Close ratios vs the prior day,
    near integer factors (k or 1/k), with min jump, min gap, and persistence.
    Returns ordered list of (date, scale) where 'scale' is the factor to multiply
    ALL prior history by to back-adjust onto the post-event scale.
    """
    o = df[f"{prefix}_open"]
    c = df[f"{prefix}_close"]
    ro = o / o.shift(1)
    rc = c / c.shift(1)

    candidates = []
    last_valid_dt = None
    for dt in ro.index[1:]:
        r_open, r_close = ro.loc[dt], rc.loc[dt]
        if pd.isna(r_open) or pd.isna(r_close): continue
        jump = max(abs(1 - r_open), abs(1 - r_close))
        if jump < SPLIT_MIN_JUMP:  # ignore small moves
            continue

        scale = None
        for k in SPLIT_FACTORS:
            # split (price drops): r ≈ 1/k
            if _near(r_open, 1/k, SPLIT_TOL) and _near(r_close, 1/k, SPLIT_TOL):
                scale = 1/k
                break
            # reverse split (price jumps): r ≈ k
            if _near(r_open, k, SPLIT_TOL) and _near(r_close, k, SPLIT_TOL):
                scale = k
                break
        if scale is None:
            continue

        # persistence: next day should NOT revert that scale
        idx = df.index.get_loc(dt)
        ok_persist = True
        for j in range(1, SPLIT_PERSIST_DAYS+1):
            if idx + j >= len(df.index): break
            dtp = df.index[idx + j]
            # day-over-day move after event should not be another near-integer undoing
            rpo = o.loc[dtp] / o.loc[df.index[idx + j - 1]]
            rpc = c.loc[dtp] / c.loc[df.index[idx + j - 1]]
            # if it flips right back near inverse scale, reject
            if _near(rpo, 1/scale, SPLIT_TOL) and _near(rpc, 1/scale, SPLIT_TOL):
                ok_persist = False
                break
        if not ok_persist:
            continue

        # min gap between events
        if last_valid_dt is not None:
            if (dt - last_valid_dt).days < SPLIT_MIN_GAP_DAYS:
                continue

        candidates.append((dt, scale))
        last_valid_dt = dt

    return candidates

def back_adjust(df: pd.DataFrame, prefix: str, events):
    """
    Apply back-adjustment: for each (date, scale), multiply ALL prior rows by 'scale'
    so the series is continuous on the *post-event* scale.
    """
    if not events: return df.copy()
    out = df.copy()
    for dt, scale in sorted(events):
        mask = out.index < dt
        out.loc[mask, [f"{prefix}_open", f"{prefix}_close"]] *= scale
    return out

def add_wm(fig): fig.text(0.01, 0.01, WATERMARK, fontsize=11, alpha=0.35, ha="left", va="bottom")
def finalize(fig): add_wm(fig); plt.tight_layout(); plt.show()

def metrics(inc_daily: pd.Series):
    inc = inc_daily.fillna(0.0).astype(float)
    equity = (1 + inc).cumprod()
    roll_max = equity.cummax()
    dd = equity / roll_max - 1.0
    cg = equity.iloc[-1] - 1.0
    wins = (inc > 0).sum(); losses = (inc < 0).sum()
    wl = (wins / losses) if losses>0 else np.nan
    var95 = -np.nanpercentile(inc.replace([np.inf,-np.inf], np.nan).dropna(), 5)
    ann_vol = inc.std(ddof=0) * np.sqrt(252)
    n = len(inc)
    cagr = equity.iloc[-1]**(252 / n) - 1 if n>0 else np.nan
    trr = (cagr / ann_vol) if (ann_vol and ann_vol != 0) else np.nan
    return equity, dd, cg, wl, var95, ann_vol, cagr, trr

# ---------------- Load data ----------------
p_adr   = resolve_path(FN_ADR)
p_local = resolve_path(FN_LOCAL)

adr_raw = load_daily_ohlc(p_adr, "ADR")
loc_raw = load_daily_ohlc(p_local, "LOCAL")

# --- Split handling ---
if ADJUST_SPLITS:
    adr_events = detect_splits_strict(adr_raw, "ADR")
    loc_events = detect_splits_strict(loc_raw, "LOCAL")
    adr = back_adjust(adr_raw, "ADR", adr_events)
    loc = back_adjust(loc_raw, "LOCAL", loc_events)
    print("ADR split events (strict):", adr_events)
    print("LOCAL split events (strict):", loc_events)
else:
    adr, loc = adr_raw.copy(), loc_raw.copy()
    adr_events, loc_events = [], []

# --- Align on intersection of dates ---
idx = adr.index.intersection(loc.index)
adr = adr.reindex(idx)
loc = loc.reindex(idx)

panel = pd.concat([adr, loc], axis=1).dropna()
print("Date span:", panel.index.min().date(), "→", panel.index.max().date(), "| obs:", len(panel))

# --- Build blackout mask (T and T+1) ---
event_days = set([dt for dt,_ in adr_events] + [dt for dt,_ in loc_events])
blackout = pd.Series(False, index=panel.index)
for dt in event_days:
    if dt in blackout.index:
        blackout.loc[dt] = True
    # T+1
    nxt = panel.index[panel.index.get_indexer([dt], method="backfill")[0] + 1] if dt in panel.index else None
    # safer: mark the next calendar row if it exists
    i = blackout.index.get_indexer([dt], method='nearest')[0]
    if i+1 < len(blackout.index):
        blackout.iloc[i+1] = True

# ---------------- Rolling OLS for z (using ADJUSTED OPENs) ----------------
exog_open = sm.add_constant(panel["LOCAL_open"])
endog_open = panel["ADR_open"]

rol = RollingOLS(endog_open, exog_open, window=ROLL_WINDOW, min_nobs=ROLL_WINDOW).fit()
alpha_t = rol.params["const"]
beta_t  = rol.params["LOCAL_open"]

resid_open = endog_open - (alpha_t + beta_t * panel["LOCAL_open"])
resid_mu = resid_open.rolling(ROLL_WINDOW).mean()
resid_sd = resid_open.rolling(ROLL_WINDOW).std(ddof=0)
z = (resid_open - resid_mu) / resid_sd

# ---------------- Daily Open→Close returns ----------------
panel["ADR_ret_day"]   = panel["ADR_close"]   / panel["ADR_open"]   - 1.0
panel["LOCAL_ret_day"] = panel["LOCAL_close"] / panel["LOCAL_open"] - 1.0

# Compose working frame
work = panel.join(alpha_t.rename("alpha")).join(beta_t.rename("beta")).join(
    resid_open.rename("resid")).join(z.rename("z")).dropna()

# ensure blackout mask matches work index
blackout = blackout.reindex(work.index).fillna(False)

# ---------------- Strategy runner with stop-loss ----------------
def run_threshold_stop(thr: float, stop_cap: float, hedge_mode: str = HEDGE_MODE):
    """
    Trade at OPEN, exit at CLOSE same day.
      z >= +thr -> ADR rich -> Short ADR, Long (β or 1x) Local
      z <= -thr -> ADR cheap -> Long ADR, Short (β or 1x) Local
      else -> flat
    Pair return:
      beta-hedged:   rich: -ADR + β·LOCAL   |   cheap: +ADR - β·LOCAL
      equal-notion:  rich: -ADR + 1·LOCAL   |   cheap: +ADR - 1·LOCAL
    Stop-loss: cap daily pair return at -stop_cap (no upside cap).
    Costs: apply 4 * TC_BPS_PER_SIDE on days we are active (same-day round trip, two legs).
    Blackout: no trades on split days T & T+1.
    """
    df = work.copy()
    zsig = df["z"]
    beta = df["beta"]

    rich  = (zsig >= +thr)
    cheap = (zsig <= -thr)
    active = (rich | cheap) & (~blackout)

    if hedge_mode.lower() == "beta":
        pair_ret = np.where(rich, -df["ADR_ret_day"] + beta*df["LOCAL_ret_day"], 0.0)
        pair_ret = np.where(cheap,  df["ADR_ret_day"] - beta*df["LOCAL_ret_day"], pair_ret)
    else:
        pair_ret = np.where(rich, -df["ADR_ret_day"] + df["LOCAL_ret_day"], 0.0)
        pair_ret = np.where(cheap,  df["ADR_ret_day"] - df["LOCAL_ret_day"], pair_ret)

    # Transaction costs for active days
    tc = (TC_BPS_PER_SIDE / 1e4) * 4.0
    after_cost = pair_ret - tc * active.astype(float)

    # Stop-loss cap (approx with OHLC)
    after_stop = np.maximum(after_cost, -stop_cap)

    # Metrics
    equity, dd, cg, wl, var95, ann_vol, cagr, trr = metrics(pd.Series(after_stop, index=df.index))

    # ---- Plots ----
    def annotate_last(ax, x, y, fmt="{:.3f}", dx=8, dy=0):
        ax.annotate(fmt.format(y[-1]), xy=(x[-1], y[-1]), xytext=(dx, dy), textcoords="offset points", va="center")

    # Z and entries
    fig1, ax1 = plt.subplots(figsize=(11.5, 5.2))
    ax1.plot(df.index, zsig, label="Residual z")
    ax1.axhline(0, linewidth=1)
    if thr>0:
        ax1.axhline(+thr, linestyle="--", alpha=0.8, label=f"+{thr}σ")
        ax1.axhline(-thr, linestyle="--", alpha=0.8, label=f"−{thr}σ")
    ax1.scatter(df.index[rich & ~blackout],  zsig[rich & ~blackout],  s=14, label="Short ADR / Long Local", alpha=0.8)
    ax1.scatter(df.index[cheap & ~blackout], zsig[cheap & ~blackout], s=14, label="Long ADR / Short Local",  alpha=0.8)
    ax1.set_title(f"Residual Z & Entries — thr={thr}σ, stop={int(stop_cap*100)}%, hedge={hedge_mode}")
    ax1.set_ylabel("Z"); ax1.set_xlabel("Date"); ax1.grid(True, alpha=0.3)
    ax1.xaxis.set_major_formatter(DATE_FMT_D); ax1.legend()
    finalize(fig1)

    # Equity
    fig2, ax2 = plt.subplots(figsize=(11.5, 5.2))
    ax2.plot(df.index, equity, label="Equity")
    ax2.set_title(f"Equity — thr={thr}σ, stop={int(stop_cap*100)}%, hedge={hedge_mode}")
    ax2.set_ylabel("Growth of $1"); ax2.set_xlabel("Date"); ax2.grid(True, alpha=0.3)
    ax2.xaxis.set_major_formatter(DATE_FMT_D); ax2.legend()
    annotate_last(ax2, df.index, equity.values, fmt="{:.3f}")
    finalize(fig2)

    # Daily returns
    fig3, ax3 = plt.subplots(figsize=(11.5, 4.2))
    colors = np.where(after_stop>=0, "tab:blue", "tab:orange")
    ax3.bar(df.index, after_stop, width=1.8, align="center", color=colors)
    ax3.axhline(0, linewidth=1)
    ax3.set_title(f"Daily Returns (post-cost, stop-capped) — thr={thr}σ, stop={int(stop_cap*100)}%")
    ax3.set_ylabel("Return"); ax3.set_xlabel("Date"); ax3.grid(True, alpha=0.25)
    ax3.xaxis.set_major_formatter(DATE_FMT_D)
    finalize(fig3)

    # Per-active-day summary (head)
    entries = df.loc[active, ["z","beta","ADR_ret_day","LOCAL_ret_day"]].copy()
    entries["Side"] = np.where(rich.loc[entries.index], "Short ADR / Long Local", "Long ADR / Short Local")
    entries["PnL_day"] = pd.Series(after_stop, index=df.index).loc[entries.index]
    entries.index.name = "Date"

    # Prints
    print(f"\n=== ADR/Local StatArb — thr={thr}σ, stop={int(stop_cap*100)}%, hedge={hedge_mode} ===")
    print(f"Obs={len(df)} | Active days={(active).sum()} | Blackout days={int(blackout.sum())}")
    print(f"Cumulative Gain: {cg:.2%}")
    print(f"Max Drawdown: {dd.min():.2%}")
    print(f"Win/Loss ratio: {wl:.3f}")
    print(f"VaR 95% (1-day): {var95:.2%}")
    print(f"Annualized Volatility: {ann_vol:.2%}")
    print(f"CAGR: {cagr:.2%}")
    print(f"Total Risk Ratio (CAGR/Ann.Vol): {trr:.3f}")
    print("\n--- Active-Day P&L (sample head) ---")
    print(entries.head(10).to_string(float_format=lambda x: f"{x:,.4f}"))

    # Save CSV
    out = pd.concat(
        [
            df[["ADR_open","ADR_close","LOCAL_open","LOCAL_close"]],
            pd.DataFrame({"alpha":df["alpha"], "beta":beta, "z":zsig}),
            pd.DataFrame({"ADR_ret_day":df["ADR_ret_day"], "LOCAL_ret_day":df["LOCAL_ret_day"]}),
            pd.Series(after_stop, name="Strat_ret"),
            pd.Series(equity, name="Equity"),
            pd.Series(dd, name="Drawdown"),
        ], axis=1
    )
    out_path = f"/content/adr_local_OC_thr_{str(thr).replace('.','p')}_stop_{int(stop_cap*100)}_{hedge_mode}.csv"
    out.to_csv(out_path)
    print(f"Saved CSV -> {out_path}")

# ---------------- Run grid: thresholds × daily stop-loss ----------------
print("β-hedged, no explicit FX. Splits strict-detected & back-adjusted; blackout T & T+1.")
for thr in Z_THRESHOLDS:
    for stop in STOP_LIST:
        run_threshold_stop(thr, stop, hedge_mode=HEDGE_MODE)


```
