```python
# ==========================================================
# ADR vs Local StatArb — Barriered Execution + OOS Winsorized Variants
# IMPUY (ADR, USD) vs IMPJ (Local) — Daily Open→Close, β-hedged
# Trade only if |expected edge| > max(cost_barrier, β-scaled round-trip cost)
# Strict split handling + blackout (T & T+1); costs and daily stop-loss caps
# Also compute out-of-sample winsorized equity for central 95% & 90% returns
# Watermark: "Orca's Capital Strategies"
# ==========================================================
import os, glob, math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import statsmodels.api as sm
from statsmodels.regression.rolling import RollingOLS

# ---------------- Config ----------------
WATERMARK = "Orca's Capital Strategies"
DATE_FMT_D = DateFormatter("%Y-%m-%d")

FN_ADR    = "IMPUY Historical Data.csv"      # ADR (USD)
FN_LOCAL  = "IMPJ Historical Data.csv"       # Local listing

ROLL_WINDOW = 60              # days for rolling z and z→edge mapping
STOP_LIST = [0.025, 0.050, 0.075]            # daily stop caps (2.5%, 5%, 7.5%)
COST_BARRIERS = [0.0030, 0.0048, 0.0060]     # 30, 48, 60 bps scenario barriers

# Execution frictions (per side, per leg) in bps
FEES_BPS_PER_SIDE       = 3
SLIPPAGE_BPS_PER_SIDE   = 2
BA_BPS_PER_SIDE_ADR     = 6
BA_BPS_PER_SIDE_LOCAL   = 8
# β-scaled daily round-trip cost (both legs; entry+exit) computed dynamically

# Splits detection (conservative)
ADJUST_SPLITS = True
SPLIT_FACTORS = [2, 3, 4, 5, 10]
SPLIT_TOL = 0.02              # ±2% around integer ratios
SPLIT_MIN_JUMP = 0.45         # require ≥45% jump day-over-day
SPLIT_MIN_GAP_DAYS = 60       # min gap between valid events on same leg
SPLIT_PERSIST_DAYS = 2        # post-event persistence ≥2 days

# ---------------- Helpers ----------------
def resolve_path(filename):
    for p in (f"/content/{filename}", f"/mnt/data/{filename}", filename):
        if os.path.isfile(p): return p
    for base in ("/content", "/mnt/data", "."):
        hits = glob.glob(os.path.join(base, "**", filename), recursive=True)
        if hits: return hits[0]
    raise FileNotFoundError(f"Could not find: {filename}")

def robust_date(s: pd.Series) -> pd.Series:
    s = s.astype(str)
    d1 = pd.to_datetime(s, dayfirst=True, errors="coerce")
    d2 = pd.to_datetime(s, dayfirst=False, errors="coerce")
    return d1 if d1.max() >= d2.max() else d2

def parse_number(series: pd.Series) -> pd.Series:
    s = series.astype(str).str.strip().replace({"-": np.nan})
    s = s.str.replace("%","",regex=False)
    smpl = s.sample(min(200, len(s)), random_state=0)
    comma_no_dot = ((smpl.str.contains(",")) & ~smpl.str.contains(r"\.")).mean()
    if comma_no_dot > 0.6:
        s = s.str.replace(".", "", regex=False)
        s = s.str.replace(",", ".", regex=False)
    else:
        s = s.str.replace(",", "", regex=False)
    return pd.to_numeric(s, errors="coerce")

def load_daily_ohlc(path: str, prefix: str) -> pd.DataFrame:
    df = pd.read_csv(path)
    df.columns = [c.strip() for c in df.columns]
    if not {"Date","Open"}.issubset(df.columns) or ("Price" not in df.columns and "Close" not in df.columns):
        raise ValueError(f"{path} must have Date, Open and Price/Close columns. Got: {df.columns.tolist()}")
    df["Date"]  = robust_date(df["Date"])
    df["Open"]  = parse_number(df["Open"])
    ccol = "Price" if "Price" in df.columns else "Close"
    df["Close"] = parse_number(df[ccol])
    out = df.dropna(subset=["Date","Open","Close"]).sort_values("Date").set_index("Date")
    out = out[~out.index.duplicated(keep="last")]
    return out.rename(columns={"Open": f"{prefix}_open", "Close": f"{prefix}_close"})[[f"{prefix}_open", f"{prefix}_close"]]

def _near(x, target, tol): return abs(x - target) <= tol

def detect_splits_strict(df: pd.DataFrame, prefix: str):
    o = df[f"{prefix}_open"]; c = df[f"{prefix}_close"]
    ro, rc = o/o.shift(1), c/c.shift(1)
    events, last_dt = [], None
    for dt in ro.index[1:]:
        r_open, r_close = ro.loc[dt], rc.loc[dt]
        if pd.isna(r_open) or pd.isna(r_close): continue
        jump = max(abs(1 - r_open), abs(1 - r_close))
        if jump < SPLIT_MIN_JUMP: continue
        scale = None
        for k in SPLIT_FACTORS:
            if _near(r_open, 1/k, SPLIT_TOL) and _near(r_close, 1/k, SPLIT_TOL): scale = 1/k; break
            if _near(r_open, k,   SPLIT_TOL) and _near(r_close, k,   SPLIT_TOL): scale = k;   break
        if scale is None: continue
        # persistence
        idx = df.index.get_loc(dt)
        ok = True
        for j in range(1, SPLIT_PERSIST_DAYS+1):
            if idx+j >= len(df.index): break
            nxt = df.index[idx+j]
            rpo = o.loc[nxt] / o.loc[df.index[idx+j-1]]
            rpc = c.loc[nxt] / c.loc[df.index[idx+j-1]]
            if _near(rpo, 1/scale, SPLIT_TOL) and _near(rpc, 1/scale, SPLIT_TOL):
                ok = False; break
        if not ok: continue
        if last_dt is not None and (dt - last_dt).days < SPLIT_MIN_GAP_DAYS: continue
        events.append((dt, scale)); last_dt = dt
    return events

def back_adjust(df: pd.DataFrame, prefix: str, events):
    if not events: return df.copy()
    out = df.copy()
    for dt, scale in sorted(events):
        mask = out.index < dt
        out.loc[mask, [f"{prefix}_open", f"{prefix}_close"]] *= scale
    return out

def add_wm(fig): fig.text(0.01, 0.01, WATERMARK, fontsize=11, alpha=0.35, ha="left", va="bottom")
def finalize(fig): add_wm(fig); plt.tight_layout(); plt.show()

def annotate_last(ax, x, y, fmt="{:.3f}", dx=8, dy=0):
    ax.annotate(fmt.format(y[-1]), xy=(x[-1], y[-1]),
                xytext=(dx, dy), textcoords="offset points", va="center")

def metrics(inc_daily: pd.Series):
    inc = inc_daily.fillna(0.0).astype(float)
    equity = (1 + inc).cumprod()
    roll_max = equity.cummax()
    dd = equity/roll_max - 1.0
    cg = equity.iloc[-1] - 1.0
    wins = (inc > 0).sum(); losses = (inc < 0).sum()
    wl = (wins / losses) if losses>0 else np.nan
    var95 = -np.nanpercentile(inc.replace([np.inf,-np.inf], np.nan).dropna(), 5)
    ann_vol = inc.std(ddof=0) * np.sqrt(252)
    n = len(inc)
    cagr = equity.iloc[-1]**(252 / n) - 1 if n>0 else np.nan
    trr = (cagr / ann_vol) if (ann_vol and ann_vol != 0) else np.nan
    return equity, dd, cg, wl, var95, ann_vol, cagr, trr

def winsorize_expanding(x: pd.Series, central: float, min_periods: int = 60) -> pd.Series:
    """
    Out-of-sample winsorization: for day t, clip x_t to [Q_lo, Q_hi] computed
    from x up to t-1. 'central' = 0.95 or 0.90 -> keep central mass.
    """
    lo = (1.0 - central) / 2.0
    hi = 1.0 - lo
    # expanding quantiles from history up to t-1 (lag via shift)
    qlo = x.shift(1).expanding(min_periods=min_periods).quantile(lo)
    qhi = x.shift(1).expanding(min_periods=min_periods).quantile(hi)
    return x.clip(lower=qlo, upper=qhi)

# ---------------- Load & adjust ----------------
p_adr, p_local = resolve_path(FN_ADR), resolve_path(FN_LOCAL)
adr_raw, loc_raw = load_daily_ohlc(p_adr, "ADR"), load_daily_ohlc(p_local, "LOCAL")

if ADJUST_SPLITS:
    adr_events = detect_splits_strict(adr_raw, "ADR")
    loc_events = detect_splits_strict(loc_raw, "LOCAL")
    adr = back_adjust(adr_raw, "ADR", adr_events)
    loc = back_adjust(loc_raw, "LOCAL", loc_events)
    print("ADR split events (strict):", adr_events)
    print("LOCAL split events (strict):", loc_events)
else:
    adr, loc = adr_raw.copy(), loc_raw.copy()
    adr_events, loc_events = [], []

idx = adr.index.intersection(loc.index)
panel = pd.concat([adr.reindex(idx), loc.reindex(idx)], axis=1).dropna()
print("Span:", panel.index.min().date(), "→", panel.index.max().date(), "| obs:", len(panel))

# Blackout T & T+1 around split events
event_days = set([dt for dt,_ in adr_events] + [dt for dt,_ in loc_events])
blackout = pd.Series(False, index=panel.index)
for dt in event_days:
    if dt in blackout.index: blackout.loc[dt] = True
    i = blackout.index.get_indexer([dt], method='nearest')[0]
    if i+1 < len(blackout): blackout.iloc[i+1] = True

# ---------------- Signal: rolling residual z on adjusted OPENs ----------------
X = sm.add_constant(panel["LOCAL_open"])
y = panel["ADR_open"]
rol = RollingOLS(y, X, window=ROLL_WINDOW, min_nobs=ROLL_WINDOW).fit()
alpha_t, beta_t = rol.params["const"], rol.params["LOCAL_open"]
resid_open = y - (alpha_t + beta_t * panel["LOCAL_open"])
res_mu = resid_open.rolling(ROLL_WINDOW).mean()
res_sd = resid_open.rolling(ROLL_WINDOW).std(ddof=0)
z = (resid_open - res_mu) / res_sd

# ---------------- Daily Open→Close returns ----------------
panel["ADR_ret_day"]   = panel["ADR_close"]   / panel["ADR_open"]   - 1.0
panel["LOCAL_ret_day"] = panel["LOCAL_close"] / panel["LOCAL_open"] - 1.0

work = panel.join(alpha_t.rename("alpha")).join(beta_t.rename("beta")).join(
    pd.DataFrame({"z": z, "resid": resid_open})
).dropna()
blackout = blackout.reindex(work.index).fillna(False)

# ---------------- β-scaled dynamic round-trip cost (entry+exit, both legs) ----------------
per_side_common = (FEES_BPS_PER_SIDE + SLIPPAGE_BPS_PER_SIDE) / 1e4
per_side_adr    = per_side_common + BA_BPS_PER_SIDE_ADR   / 1e4
per_side_local  = per_side_common + BA_BPS_PER_SIDE_LOCAL / 1e4
def cost_per_day(beta):
    # beta scales the local leg notional
    return 2.0 * (per_side_adr + abs(beta) * per_side_local)

# ---------------- Rolling mapping: expected edge from z (lagged coefficients) ----------------
# Define long-cheap pair return (what you earn if z<0 and you go Long ADR / Short Local)
ret_LC = work["ADR_ret_day"] - work["beta"] * work["LOCAL_ret_day"]
ols_edge = RollingOLS(ret_LC, sm.add_constant(work["z"]), window=ROLL_WINDOW, min_nobs=ROLL_WINDOW).fit()
a_hat = ols_edge.params["const"].shift(1)  # lag to avoid same-day look-ahead
b_hat = ols_edge.params["z"].shift(1)

def run_barrier(cost_barrier: float, stop_cap: float):
    df = work.copy()
    beta = df["beta"]; zsig = df["z"]

    # Predicted edge for the *actual* side we would take
    pred_lc = a_hat + b_hat * zsig            # predicted long-cheap edge
    pred_pos = np.where(zsig <= 0, pred_lc, -pred_lc)  # if rich, flip sign

    # Dynamic β-scaled cost & needed hurdle
    dyn_cost = cost_per_day(beta)
    need_edge = np.maximum(cost_barrier, dyn_cost)

    # Entry rule (blackout days excluded)
    enter = (np.abs(pred_pos) > need_edge) & (~blackout)

    # Realized pair return based on side taken
    pair_ret = np.where(zsig > 0, -df["ADR_ret_day"] + beta * df["LOCAL_ret_day"],
                                   df["ADR_ret_day"] - beta * df["LOCAL_ret_day"])

    # Subtract actual dynamic cost on active days
    after_cost = pair_ret - dyn_cost * enter.astype(float)

    # Apply daily stop-loss cap (no upside cap)
    realized = np.maximum(after_cost, -stop_cap)
    realized = pd.Series(realized, index=df.index)

    # ---------------- Baseline metrics ----------------
    equity, dd, cg, wl, var95, ann_vol, cagr, trr = metrics(realized)

    # ---------------- Out-of-sample winsorized variants ----------------
    realized_w95 = winsorize_expanding(realized, central=0.95, min_periods=ROLL_WINDOW)
    realized_w90 = winsorize_expanding(realized, central=0.90, min_periods=ROLL_WINDOW)

    eq95, dd95, cg95, wl95, var95_95, annvol95, cagr95, trr95 = metrics(realized_w95)
    eq90, dd90, cg90, wl90, var95_90, annvol90, cagr90, trr90 = metrics(realized_w90)

    # ---------------- Diagnostics ----------------
    avg_pred_traded = np.nanmean(np.abs(pred_pos[enter]))
    avg_hurdle_traded = np.nanmean(need_edge[enter])
    blocked = (~enter).sum()
    print(f"\n=== Barriered Strategy — barrier {int(cost_barrier*1e4)} bps, stop {int(stop_cap*100)}% ===")
    print(f"Obs={len(df)} | Traded days={int(enter.sum())} | Blocked days={int(blocked)} | Blackout={int(blackout.sum())}")
    print(f"Avg |pred edge| (traded): {avg_pred_traded:.3%}  |  Avg hurdle (traded): {avg_hurdle_traded:.3%}")
    print(f"[Baseline]   CumGain: {cg:.2%} | MaxDD: {dd.min():.2%} | W/L: {wl:.3f} | Ann.Vol: {ann_vol:.2%} | CAGR: {cagr:.2%} | TRR: {trr:.3f}")
    print(f"[Central95]  CumGain: {cg95:.2%} | MaxDD: {dd95.min():.2%} | W/L: {wl95:.3f} | Ann.Vol: {annvol95:.2%} | CAGR: {cagr95:.2%} | TRR: {trr95:.3f}")
    print(f"[Central90]  CumGain: {cg90:.2%} | MaxDD: {dd90.min():.2%} | W/L: {wl90:.3f} | Ann.Vol: {annvol90:.2%} | CAGR: {cagr90:.2%} | TRR: {trr90:.3f}")

    # ---------------- Plots ----------------
    # Residual Z and entries
    fig1, ax1 = plt.subplots(figsize=(11.5, 5.2))
    ax1.plot(df.index, zsig, label="Residual z")
    ax1.axhline(0, linewidth=1)
    ax1.scatter(df.index[enter & (zsig > 0)], zsig[enter & (zsig > 0)], s=12, label="Entry: Short ADR / Long Local")
    ax1.scatter(df.index[enter & (zsig < 0)], zsig[enter & (zsig < 0)], s=12, label="Entry: Long ADR / Short Local")
    ax1.set_title(f"Residual Z & Entries — barrier={int(cost_barrier*1e4)}bps, stop={int(stop_cap*100)}%")
    ax1.set_ylabel("Z"); ax1.set_xlabel("Date"); ax1.grid(True, alpha=0.3)
    ax1.xaxis.set_major_formatter(DATE_FMT_D); ax1.legend()
    finalize(fig1)

    # Equity curves: baseline + winsorized overlays
    fig2, ax2 = plt.subplots(figsize=(11.5, 5.2))
    eq = (1 + realized.fillna(0)).cumprod()
    ax2.plot(df.index, eq,   label="Baseline equity")
    ax2.plot(df.index, eq95, label="Equity — central 95% (OOS winsorized)")
    ax2.plot(df.index, eq90, label="Equity — central 90% (OOS winsorized)")
    ax2.set_title(f"Equity (baseline vs winsorized) — barrier={int(cost_barrier*1e4)}bps, stop={int(stop_cap*100)}%")
    ax2.set_ylabel("Growth of $1"); ax2.set_xlabel("Date"); ax2.grid(True, alpha=0.3)
    ax2.xaxis.set_major_formatter(DATE_FMT_D); ax2.legend()
    annotate_last(ax2, df.index, eq.values, fmt="{:.3f}")
    finalize(fig2)

    # Daily returns bars (baseline)
    fig3, ax3 = plt.subplots(figsize=(11.5, 4.2))
    colors = np.where(realized>=0, "tab:blue", "tab:orange")
    ax3.bar(df.index, realized, width=1.8, align="center", color=colors)
    ax3.axhline(0, linewidth=1)
    ax3.set_title("Daily Returns — post-cost, stop-capped (baseline)")
    ax3.set_ylabel("Return"); ax3.set_xlabel("Date"); ax3.grid(True, alpha=0.25)
    ax3.xaxis.set_major_formatter(DATE_FMT_D)
    finalize(fig3)

    # ---------------- Save CSVs ----------------
    out = pd.concat(
        [
            df[["ADR_open","ADR_close","LOCAL_open","LOCAL_close"]],
            pd.DataFrame({
                "alpha": df["alpha"], "beta": beta, "z": zsig,
                "pred_pos": pred_pos, "need_edge": need_edge, "enter": enter.astype(int)
            }),
            pd.DataFrame({"ADR_ret_day": df["ADR_ret_day"], "LOCAL_ret_day": df["LOCAL_ret_day"]}),
            realized.rename("Strat_ret_baseline"),
            realized_w95.rename("Strat_ret_capped95"),
            realized_w90.rename("Strat_ret_capped90"),
        ],
        axis=1
    )
    out_path = f"/content/adr_local_barrier_{int(cost_barrier*1e4)}bps_stop_{int(stop_cap*100)}.csv"
    out.to_csv(out_path)
    print(f"Saved CSV -> {out_path}")

# ---------------- Run scenario grid ----------------
print("Barriered execution: trade only when |predicted edge| > max(scenario barrier, β-scaled cost).")
print("Strict splits + blackout (T & T+1), daily stop caps, and out-of-sample winsorized variants included.")
for cb in COST_BARRIERS:
    for stop in STOP_LIST:
        run_barrier(cb, stop)





```
