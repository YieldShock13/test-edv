```python
# ======================================================
# HY → IG transmission (DAILY, LAGS-ONLY, +1.5σ shocks)
# Outputs focused for publication:
#   • β(h) path (h=1..30): coef, HAC se/t/p, N
#   • β(h) plot with 95% HAC CI
#   • Distributed-lag OLS (k=1..30) bar plot + HAC coef/t/p table (no R²)
#   • Granger p-values summarized at horizons 5/10/20/30 (HY→IG and IG→HY)
#   • Event-study path (±30d) for context
# Units: ΔOAS in basis points (bps) everywhere
# ======================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.stats.anova import anova_lm
from statsmodels.tsa.stattools import grangercausalitytests

# ---------- Params ----------
FILES = {
    "HY": "/content/BAMLH0A0HYM2 (9).csv",   # High Yield OAS
    "IG": "/content/BAMLC0A0CM (2).csv",     # Investment Grade OAS
}
SIGMA = 1.5
H = 30
HAC_MAXLAGS = 5
GRANGER_CHECKPOINTS = [5, 10, 20, 30]  # horizons to report p-values (HY→IG)

# ---------- Robust CSV loader ----------
def load_oas_csv(path, name):
    df = pd.read_csv(path)
    dcols = [c for c in df.columns if "date" in c.lower()]
    if not dcols:
        raise ValueError(f"No date column in {path}")
    dcol = dcols[0]
    df[dcol] = pd.to_datetime(df[dcol])
    df = df.sort_values(dcol).set_index(dcol)

    candidates = []
    for c in df.columns:
        ser = pd.to_numeric(df[c].astype(str).str.replace(",",""), errors="coerce")
        if ser.notna().sum() > 0:
            df[c] = ser
            candidates.append((c, ser.notna().sum()))
    if not candidates:
        raise ValueError(f"No numeric column in {path}")
    vcol = max(candidates, key=lambda x: x[1])[0]
    return df[[vcol]].rename(columns={vcol: name})

# ---------- Load & prepare daily ΔOAS in bps ----------
hy = load_oas_csv(FILES["HY"], "HY_OAS")
ig = load_oas_csv(FILES["IG"], "IG_OAS")

def to_bps(s):
    med = s.median()
    if pd.notna(med) and 0 < med < 20:  # looks like %
        return s * 100.0
    return s

hy = to_bps(hy["HY_OAS"]).rename("HY_OAS").to_frame()
ig = to_bps(ig["IG_OAS"]).rename("IG_OAS").to_frame()

# Align to business days and ffill
hy_b = hy.asfreq("B").ffill()
ig_b = ig.asfreq("B").ffill()

# Daily changes (bps)
dhy = hy_b.diff().rename(columns={"HY_OAS": "dHY"})
dig = ig_b.diff().rename(columns={"IG_OAS": "dIG"})

df = pd.concat([dhy, dig], axis=1).dropna().copy()

# ---------- HY +1.5σ shock flag & size (widenings only) ----------
thr = df["dHY"].mean() + SIGMA * df["dHY"].std()
df["HY_ShockFlag"] = (df["dHY"] > thr).astype(int)
df["HY_ShockSize"] = df["dHY"].where(df["HY_ShockFlag"] == 1, 0.0)

print(f"HY +{SIGMA:.1f}σ threshold (bps): {thr:.3f}")
print(f"Number of HY shock days: {int(df['HY_ShockFlag'].sum())}")

# ======================================================
# A) β(h): ΔIG_{t+h} ~ β(h)*ShockSize_t, h = 1..H (lags-only)
# ======================================================
def beta_path_lags_only(S, y, h_start=1, H=30, hac_lags=5):
    rows = []
    for h in range(h_start, H+1):
        tmp = pd.DataFrame({"y": y.shift(-h), "S": S}).dropna()
        X = sm.add_constant(tmp["S"])
        hac = sm.OLS(tmp["y"], X).fit(cov_type="HAC", cov_kwds={"maxlags": hac_lags})
        rows.append({
            "h": h,
            "beta_bps_per_1bp": float(hac.params["S"]),
            "se_hac": float(hac.bse["S"]),
            "t_hac": float(hac.tvalues["S"]),
            "p_hac": float(hac.pvalues["S"]),
            "N": int(hac.nobs),
        })
    out = pd.DataFrame(rows).set_index("h")
    sig = out.index[out["p_hac"] < 0.05]
    horizon = int(sig.max()) if len(sig) else 0
    return out, horizon

beta_tbl, horizon = beta_path_lags_only(df["HY_ShockSize"], df["dIG"], h_start=1, H=H, hac_lags=HAC_MAXLAGS)

# ======================================================
# B) Granger summaries at checkpoints (HY→IG and IG→HY)
# ======================================================
def granger_pvals(yx_df, maxlag):
    # returns p-value of ssr_ftest for the given maxlag
    res = grangercausalitytests(yx_df.values, maxlag=maxlag, verbose=False)
    # statsmodels returns dict keyed by lag; pull ssr_ftest p-value at 'maxlag'
    p = res[maxlag][0]["ssr_ftest"][1]
    return float(p)

# HY → IG: y=ΔIG, x=ΔHY
yx1 = df[["dIG","dHY"]].dropna()
# IG → HY: y=ΔHY, x=ΔIG
yx2 = df[["dHY","dIG"]].dropna()

g_summary = {}
for L in GRANGER_CHECKPOINTS:
    g_summary[f"G_p(HY→IG,≤{L})"] = granger_pvals(yx1, L)
    g_summary[f"G_p(IG→HY,≤{L})"] = granger_pvals(yx2, L)

# Attach HY→IG Granger p-values as constant columns to the beta table (compact citation in one place)
for L in GRANGER_CHECKPOINTS:
    beta_tbl[f"G_p(HY→IG,≤{L})"] = g_summary[f"G_p(HY→IG,≤{L})"]

print("\n=== β(h): ΔIG_{t+h} on HY ShockSize_t (h=1..30), lags-only ===")
print(beta_tbl.to_string())
print(f"\nTransmission horizon (last h with p<0.05): {horizon} day(s)")

print("\n=== Granger summaries (report these in text/tables) ===")
for k, v in g_summary.items():
    print(f"{k}: p = {v:.3e}")

# Plot β(h) with 95% HAC CI
ci_lo = beta_tbl["beta_bps_per_1bp"] - 1.96 * beta_tbl["se_hac"]
ci_hi = beta_tbl["beta_bps_per_1bp"] + 1.96 * beta_tbl["se_hac"]
plt.figure(figsize=(10,5))
beta_tbl["beta_bps_per_1bp"].plot(marker="o", title=f"β(h): IG response to HY +{SIGMA}σ shock (h=1..{H})")
plt.fill_between(beta_tbl.index.values, ci_lo.values, ci_hi.values, alpha=0.2)
plt.axhline(0, ls="--", c="k"); plt.xlabel("Days after HY shock (h)")
plt.ylabel("IG Δbps per 1 HY bp (shock size)"); plt.show()

# ======================================================
# C) Distributed-lag OLS: ΔIG_t ~ Σ_{k=1..30} ΔHY_{t-k} (lags-only)
#    — print HAC coef/t/p (no R²) and bar plot
# ======================================================
def build_lags(x, L, start=1):
    return pd.concat({f"x_lag{k}": x.shift(k) for k in range(start, L+1)}, axis=1)

Xlags = build_lags(df["dHY"], H, start=1)
df_dl = pd.concat([df["dIG"].rename("y"), Xlags], axis=1).dropna()

X = sm.add_constant(df_dl[[f"x_lag{k}" for k in range(1, H+1)]])
hac_dl = sm.OLS(df_dl["y"], X).fit(cov_type="HAC", cov_kwds={"maxlags": HAC_MAXLAGS})
coefs = hac_dl.params.drop("const")
tvals = hac_dl.tvalues.drop("const")
pvals = hac_dl.pvalues.drop("const")
coef_table = pd.DataFrame({"coef_bps_per_1bp": coefs, "t_hac": tvals, "p_hac": pvals})

print("\n--- Distributed-lag HAC coefs (ΔIG on ΔHY lags k=1..30) ---")
print(coef_table.to_string())

plt.figure(figsize=(12,5))
coefs.plot(kind="bar", title="ΔIG_t ~ ΔHY_{t−k} (k=1..30) — coefficients (IG bps per 1 HY bp)")
plt.axhline(0, ls="--", c="k"); plt.ylabel("IG bps per 1 HY bp"); plt.show()

# ======================================================
# D) Event study (±30d) around HY +1.5σ shock days — context only
# ======================================================
def event_study_path(y, shock_flag, window=30):
    idx = y.index
    events = idx[shock_flag==1]
    paths = []
    for t in events:
        pos = idx.get_loc(t)
        if pos-window < 0 or pos+window >= len(idx):
            continue
        seg = y.iloc[pos-window:pos+window+1].reset_index(drop=True)
        seg.index = range(-window, window+1)
        paths.append(seg)
    if not paths:
        return None
    return pd.concat(paths, axis=1).mean(axis=1)

avg_path = event_study_path(df["dIG"], df["HY_ShockFlag"], window=30)
print(f"\nHY shock count used in event study: {int(df['HY_ShockFlag'].sum())}")
if avg_path is not None:
    plt.figure(figsize=(10,5))
    avg_path.plot(title=f"Event study: Avg ΔIG around HY +{SIGMA}σ shock (±30 trading days)")
    plt.axvline(0, color="r", ls="--"); plt.xlabel("Days relative to HY shock")
    plt.ylabel("ΔIG (bps)"); plt.show()
else:
    print("(Not enough shocks to compute event study path.)")

```
