```python
import statsmodels.api as sm
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ============================================================
# 1. HALF-LIFE ESTIMATION FUNCTION
# ============================================================

def estimate_half_life(series, label):
    """
    Estimates OU half-life from:
    ΔX_t = α + β X_{t-1} + ε_t
    """
    x = series.dropna().copy()
    dx = x.diff().dropna()
    x_lag = x.shift(1).dropna()

    df = pd.concat([dx, x_lag], axis=1).dropna()
    df.columns = ["dx", "x_lag"]

    X = sm.add_constant(df["x_lag"])
    model = sm.OLS(df["dx"], X).fit()

    beta = model.params["x_lag"]

    if beta >= 0:
        half_life = np.inf
    else:
        half_life = -np.log(2) / beta

    print("\n==============================")
    print(f"HALF-LIFE ESTIMATION: {label}")
    print("==============================")
    print("Beta:", round(beta,6))
    print("t-stat:", round(model.tvalues["x_lag"],3))
    print("R²:", round(model.rsquared,4))
    print("Half-life (periods):", round(half_life,2))

    return half_life, model

# ============================================================
# 2. CONDITIONAL FORWARD RETURNS
# ============================================================

def conditional_forward_returns(price_series, z_series, horizon, threshold=-2):
    """
    Forward return when Z <= threshold
    """
    price = price_series.dropna()
    z = z_series.reindex(price.index)

    fwd_ret = price.shift(-horizon) / price - 1

    mask = z <= threshold
    conditional = fwd_ret[mask]

    print(f"\nForward {horizon}-period returns when Z <= {threshold}")
    print("Observations:", conditional.count())
    print("Mean:", round(conditional.mean(),4))
    print("Median:", round(conditional.median(),4))
    print("Std:", round(conditional.std(),4))
    print("Hit Rate:", round((conditional > 0).mean(),4))

    plt.figure(figsize=(8,4))
    plt.hist(conditional.dropna(), bins=30)
    plt.title(f"Forward {horizon} Distribution | Z <= {threshold}")
    plt.show()

    return conditional

# ============================================================
# 3. TIME-TO-MEAN-REVERSION DISTRIBUTION
# ============================================================

def time_to_reversion(z_series, threshold=2):
    """
    Measures time until Z crosses zero after |Z| >= threshold
    """
    z = z_series.dropna()
    times = []

    for i in range(len(z)):
        if abs(z.iloc[i]) >= threshold:
            start_sign = np.sign(z.iloc[i])
            for j in range(i+1, len(z)):
                if np.sign(z.iloc[j]) != start_sign:
                    times.append(j - i)
                    break

    times = pd.Series(times)

    print("\nTime-to-Reversion | |Z| >=", threshold)
    print("Observations:", len(times))
    print("Mean:", round(times.mean(),2))
    print("Median:", round(times.median(),2))
    print("Std:", round(times.std(),2))

    plt.figure(figsize=(8,4))
    plt.hist(times, bins=30)
    plt.title(f"Time to Reversion | |Z| >= {threshold}")
    plt.show()

    return times

# ============================================================
# 4. RUN FOR ALL FOUR SERIES
# ============================================================

# ---- WEEKLY SELF ----
hl_w_self, model_w_self = estimate_half_life(log_truj_w, "Weekly TRUJ Self")
conditional_forward_returns(truj, z_self_52w, 4)
time_to_reversion(z_self_52w)

# ---- WEEKLY RELATIVE ----
hl_w_rel, model_w_rel = estimate_half_life(log_rel_w, "Weekly TRUJ vs ALSI")
conditional_forward_returns(truj, z_rel_52w, 4)
time_to_reversion(z_rel_52w)

# ---- MONTHLY SELF ----
hl_m_self, model_m_self = estimate_half_life(log_truj_m, "Monthly TRUJ Self")
conditional_forward_returns(truj_m, z_self_12m, 2)
time_to_reversion(z_self_12m)

# ---- MONTHLY RELATIVE ----
hl_m_rel, model_m_rel = estimate_half_life(log_rel_m, "Monthly TRUJ vs ALSI")
conditional_forward_returns(truj_m, z_rel_12m, 2)
time_to_reversion(z_rel_12m)



```
