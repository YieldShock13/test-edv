```python
# ==========================================================
# Monthly (1st) DE vs US — Yields, Futures & Two Strategies
# Strategy A: Single-signal (Futures z) EXACTLY as your original rule
# Strategy B: Yield-only z rule at ±1σ (ignores futures z)
# ==========================================================
import os, glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import statsmodels.api as sm

# ---------------- Config ----------------
WATERMARK = "Orca's Capital Strategies"
DATE_FMT_M = DateFormatter("%Y-%m")
# For Strategy A only: which futures z to use as the signal
FUT_SIGNAL_SOURCE = "1x1"   # "1x1" uses z of (DE−US); set "hedged" to use z of (DE−β·US)

# Exact filenames
FN_DE_YLD = "Germany 2-Year Bond Yield Historical Data (2).csv"
FN_US_YLD = "United States 2-Year Bond Yield Historical Data (1).csv"
FN_DE_FUT = "Euro SCHATZ Futures Historical Data.csv"
FN_US_FUT = "US 2 Year T-Note Futures Historical Data.csv"

# ---------------- Helpers ----------------
def resolve_path(filename):
    for p in (f"/content/{filename}", f"/mnt/data/{filename}", filename):
        if os.path.isfile(p): return p
    for base in ("/content", "/mnt/data", "."):
        hits = glob.glob(os.path.join(base, "**", filename), recursive=True)
        if hits: return hits[0]
    raise FileNotFoundError(f"Could not find: {filename}")

def robust_date(s: pd.Series) -> pd.Series:
    s = s.astype(str)
    d1 = pd.to_datetime(s, dayfirst=True,  errors="coerce")
    d2 = pd.to_datetime(s, dayfirst=False, errors="coerce")
    return d1 if d1.max() >= d2.max() else d2

def parse_number(series: pd.Series) -> pd.Series:
    s = series.astype(str).str.strip().str.replace("%","",regex=False)
    smpl = s.sample(min(200, len(s)), random_state=0)
    comma_no_dot = ((smpl.str.contains(",")) & ~smpl.str.contains(r"\.")).mean()
    if comma_no_dot > 0.6:
        s = s.str.replace(".", "", regex=False)   # thousands dot
        s = s.str.replace(",", ".", regex=False)  # decimal comma → dot
    else:
        s = s.str.replace(",", "", regex=False)   # thousands comma
    return pd.to_numeric(s, errors="coerce")

def load_monthly_first(path: str, colname: str) -> pd.Series:
    df = pd.read_csv(path)
    df.columns = [c.strip() for c in df.columns]
    if "Date" not in df.columns or "Price" not in df.columns:
        raise ValueError(f"'Date' and 'Price' required in {path}. Found: {df.columns.tolist()}")
    df["Date"]  = robust_date(df["Date"])
    df["Price"] = parse_number(df["Price"])
    df = df.dropna(subset=["Date","Price"]).sort_values("Date").set_index("Date")
    # enforce one obs per month (your files are already stamped on the 1st)
    df = df[~df.index.duplicated(keep="first")]
    return df["Price"].rename(colname)

def add_wm(fig): fig.text(0.01, 0.01, WATERMARK, fontsize=12, alpha=0.35, ha="left", va="bottom")
def annotate_last(ax, x, y, fmt="{:.2f}", dx=8, dy=0): ax.annotate(fmt.format(y[-1]), xy=(x[-1], y[-1]), xytext=(dx, dy), textcoords="offset points", va="center")
def finalize(fig): add_wm(fig); plt.tight_layout(); plt.show()

def compute_metrics(inc: pd.Series):
    equity = (1 + inc).cumprod()
    roll_max = equity.cummax()
    drawdown = equity / roll_max - 1.0
    cum_gain = equity.iloc[-1] - 1.0
    wins = (inc > 0).sum(); losses = (inc < 0).sum()
    win_loss_ratio = (wins / losses) if losses > 0 else np.nan
    var95 = -np.nanpercentile(inc.dropna(), 5)  # 1-month VaR(95%)
    ann_vol = inc.std(ddof=0) * np.sqrt(12)
    n = len(inc)
    cagr = equity.iloc[-1]**(12 / n) - 1 if n > 0 else np.nan
    trr = (cagr / ann_vol) if (ann_vol and ann_vol != 0) else np.nan
    return {
        "equity": equity, "drawdown": drawdown, "cum_gain": cum_gain, "wins": wins, "losses": losses,
        "win_loss_ratio": win_loss_ratio, "var95": var95, "ann_vol": ann_vol, "cagr": cagr, "trr": trr
    }

# ---------------- Load data ----------------
p_de_yld = resolve_path(FN_DE_YLD)
p_us_yld = resolve_path(FN_US_YLD)
p_de_fut = resolve_path(FN_DE_FUT)
p_us_fut = resolve_path(FN_US_FUT)

DE2Y = load_monthly_first(p_de_yld, "DE2Y_yld")
US2Y = load_monthly_first(p_us_yld, "US2Y_yld")
DE_F = load_monthly_first(p_de_fut, "DE_fut")
US_F = load_monthly_first(p_us_fut, "US_fut")

# Align indices
ix_yld = DE2Y.index.intersection(US2Y.index)
yld = pd.concat([DE2Y.reindex(ix_yld), US2Y.reindex(ix_yld)], axis=1).dropna()

ix_fut = DE_F.index.intersection(US_F.index)
fut = pd.concat([DE_F.reindex(ix_fut), US_F.reindex(ix_fut)], axis=1).dropna()

# ---------------- Yield spread (bps) + z + OLS ----------------
yld["Spread_bps"] = (yld["DE2Y_yld"] - yld["US2Y_yld"]) * 100.0
mu_y, sd_y = yld["Spread_bps"].mean(), yld["Spread_bps"].std(ddof=0)
yld["Z_yield"] = (yld["Spread_bps"] - mu_y) / sd_y if sd_y != 0 else np.nan

# OLS: DE2Y_yld ~ const + beta * US2Y_yld (kept/printed later)
Xy = sm.add_constant(yld["US2Y_yld"])
yy = yld["DE2Y_yld"]
ols_yield = sm.OLS(yy, Xy).fit()
beta_y = float(ols_yield.params["US2Y_yld"]); alpha_y = float(ols_yield.params["const"]); r2_y = float(ols_yield.rsquared)

# ---------------- Futures spreads + z ----------------
# 1×1
fut["Spread_1x1"] = fut["DE_fut"] - fut["US_fut"]
mu_1x1, sd_1x1 = fut["Spread_1x1"].mean(), fut["Spread_1x1"].std(ddof=0)
fut["Z_1x1"] = (fut["Spread_1x1"] - mu_1x1) / sd_1x1 if sd_1x1 != 0 else np.nan

# OLS-hedged (DE_fut ~ const + beta * US_fut)
Xf = sm.add_constant(fut["US_fut"])
yf = fut["DE_fut"]
ols_fut = sm.OLS(yf, Xf).fit()
beta_f = float(ols_fut.params["US_fut"]); alpha_f = float(ols_fut.params["const"])

fut["Spread_hedged"] = fut["DE_fut"] - beta_f * fut["US_fut"]
mu_h, sd_h = fut["Spread_hedged"].mean(), fut["Spread_hedged"].std(ddof=0)
fut["Z_hedged"] = (fut["Spread_hedged"] - mu_h) / sd_h if sd_h != 0 else np.nan

# ---------------- Forward returns (signal at t, hold to t+1) ----------------
DE_ret_fwd = DE_F.shift(-1) / DE_F - 1.0
US_ret_fwd = US_F.shift(-1) / US_F - 1.0
ret_df = pd.concat([DE_ret_fwd.rename("DE_ret_fwd"), US_ret_fwd.rename("US_ret_fwd")], axis=1)

# ==========================================================
# STRATEGY A — EXACT original single-signal rule (futures z)
# ==========================================================
panelA = fut.join(ret_df, how="inner").dropna()
zsigA = panelA["Z_1x1"] if FUT_SIGNAL_SOURCE.lower() == "1x1" else panelA["Z_hedged"]

long_DE_A = (zsigA < 0)
long_US_A = (zsigA > 0)

panelA["Strat_ret"] = 0.0
panelA.loc[long_DE_A, "Strat_ret"] = panelA.loc[long_DE_A, "DE_ret_fwd"]
panelA.loc[long_US_A, "Strat_ret"] = panelA.loc[long_US_A, "US_ret_fwd"]

tradesA = panelA.loc[long_DE_A | long_US_A, ["DE_ret_fwd","US_ret_fwd","Strat_ret"]].copy()
tradesA["Leg"] = np.where(long_DE_A.loc[tradesA.index], "DE", "US")

metricsA = compute_metrics(panelA["Strat_ret"])
equityA, drawdownA = metricsA["equity"], metricsA["drawdown"]

# ---------------- Plots for Strategy A ----------------
# Yield spread & z (visuals retained)
figA1, axA1 = plt.subplots(figsize=(11,5.0))
axA1.plot(yld.index, yld["Spread_bps"])
axA1.set_title("Yield Spread (DE2Y − US2Y) — Monthly (1st)")
axA1.set_ylabel("bps"); axA1.set_xlabel("Date"); axA1.grid(True, alpha=0.3)
axA1.xaxis.set_major_formatter(DATE_FMT_M)
annotate_last(axA1, yld.index, yld["Spread_bps"].values, fmt="{:.1f} bps")
finalize(figA1)

figA2, axA2 = plt.subplots(figsize=(11,5.0))
axA2.plot(yld.index, yld["Z_yield"], label="Yield z")
axA2.axhline(0, linewidth=1)
axA2.set_title("Z-Score of Yield Spread (DE2Y − US2Y)")
axA2.set_ylabel("Z"); axA2.set_xlabel("Date"); axA2.grid(True, alpha=0.3)
axA2.xaxis.set_major_formatter(DATE_FMT_M)
annotate_last(axA2, yld.index, yld["Z_yield"].values, fmt="{:.2f}")
axA2.legend(); finalize(figA2)

# Futures z visuals (both shown)
figA3, axA3 = plt.subplots(figsize=(11,5.0))
axA3.plot(fut.index, fut["Z_1x1"], label="Futures z (1×1)")
axA3.plot(fut.index, fut["Z_hedged"], label=f"Futures z (hedged β={beta_f:.3f})")
axA3.axhline(0, linewidth=1)
axA3.set_title("Futures Z-Scores — Monthly (1st)")
axA3.set_ylabel("Z"); axA3.set_xlabel("Date"); axA3.grid(True, alpha=0.3)
axA3.xaxis.set_major_formatter(DATE_FMT_M)
annotate_last(axA3, fut.index, fut["Z_1x1"].values, fmt="{:.2f}")
annotate_last(axA3, fut.index, fut["Z_hedged"].values, fmt="{:.2f}", dy=-12)
axA3.legend(); finalize(figA3)

# Futures prices with ENTRY markers for A
figA4, axA4 = plt.subplots(figsize=(12,5.6))
axA4.plot(fut.index, fut["DE_fut"], label="DE futures")
axA4.plot(fut.index, fut["US_fut"], label="US futures")
de_entries_A = panelA.index[long_DE_A]; us_entries_A = panelA.index[long_US_A]
axA4.scatter(de_entries_A, fut.loc[de_entries_A, "DE_fut"], marker="^", s=90, label="ENTRY: Long DE", zorder=5)
axA4.scatter(us_entries_A, fut.loc[us_entries_A, "US_fut"], marker="v", s=90, label="ENTRY: Long US", zorder=5)
axA4.set_title(f"Strategy A Entries — FutZ source: {FUT_SIGNAL_SOURCE.upper()}")
axA4.set_ylabel("Price"); axA4.set_xlabel("Date"); axA4.grid(True, alpha=0.3)
axA4.xaxis.set_major_formatter(DATE_FMT_M)
annotate_last(axA4, fut.index, fut["DE_fut"].values, fmt="{:.2f}")
annotate_last(axA4, fut.index, fut["US_fut"].values, fmt="{:.2f}")
axA4.legend(); finalize(figA4)

# Equity, returns, drawdown for A
figA5, axA5 = plt.subplots(figsize=(11,5.0))
axA5.plot(panelA.index, equityA, label="Equity")
axA5.set_title("Strategy A — Equity Curve (Monthly, 1st)")
axA5.set_ylabel("Growth of $1"); axA5.set_xlabel("Date"); axA5.grid(True, alpha=0.3)
axA5.xaxis.set_major_formatter(DATE_FMT_M)
annotate_last(axA5, panelA.index, equityA.values, fmt="{:.3f}")
axA5.legend(); finalize(figA5)

figA6, axA6 = plt.subplots(figsize=(11,4.2))
colorsA = np.where(panelA["Strat_ret"] >= 0, "tab:blue", "tab:orange")
axA6.bar(panelA.index, panelA["Strat_ret"].values, width=20, align="center", color=colorsA)
axA6.axhline(0, linewidth=1)
axA6.set_title("Strategy A — Monthly Returns (Wins blue / Losses orange)")
axA6.set_ylabel("Return"); axA6.set_xlabel("Date"); axA6.grid(True, alpha=0.25)
axA6.xaxis.set_major_formatter(DATE_FMT_M)
finalize(figA6)

figA7, axA7 = plt.subplots(figsize=(11,3.8))
axA7.plot(panelA.index, drawdownA)
axA7.set_title("Strategy A — Drawdown")
axA7.set_ylabel("Drawdown"); axA7.set_xlabel("Date"); axA7.grid(True, alpha=0.3)
axA7.xaxis.set_major_formatter(DATE_FMT_M)
finalize(figA7)

# Yield OLS scatter with fit (retained)
figAy, axAy = plt.subplots(figsize=(6.6,6.0))
axAy.scatter(yld["US2Y_yld"], yld["DE2Y_yld"], s=30, alpha=0.7)
x_grid = np.linspace(yld["US2Y_yld"].min(), yld["US2Y_yld"].max(), 50)
axAy.plot(x_grid, alpha_y + beta_y * x_grid, linewidth=2)
axAy.set_title(f"Yield OLS: DE2Y = α + β·US2Y  (β={beta_y:.3f}, α={alpha_y:.3f}, R²={r2_y:.3f})")
axAy.set_xlabel("US 2Y yield"); axAy.set_ylabel("DE 2Y yield"); axAy.grid(True, alpha=0.3)
add_wm(figAy); plt.tight_layout(); plt.show()

# ==========================================================
# STRATEGY B — Yield-only rule at ±1σ (ignores futures z)
# ==========================================================
panelB = fut.join(yld[["Z_yield"]], how="inner").join(ret_df, how="inner").dropna()

# Enter when |Z_yield| >= 1:
long_DE_B = (panelB["Z_yield"] <= -1.0)   # more negative than −1σ
long_US_B = (panelB["Z_yield"] >= +1.0)   # more positive than +1σ

panelB["Strat_ret"] = 0.0
panelB.loc[long_DE_B, "Strat_ret"] = panelB.loc[long_DE_B, "DE_ret_fwd"]
panelB.loc[long_US_B, "Strat_ret"] = panelB.loc[long_US_B, "US_ret_fwd"]

tradesB = panelB.loc[long_DE_B | long_US_B, ["DE_ret_fwd","US_ret_fwd","Strat_ret","Z_yield"]].copy()
tradesB["Leg"] = np.where(long_DE_B.loc[tradesB.index], "DE", "US")

metricsB = compute_metrics(panelB["Strat_ret"])
equityB, drawdownB = metricsB["equity"], metricsB["drawdown"]

# ---------------- Plots for Strategy B ----------------
# Futures prices with ENTRY markers for B (driven by Z_yield)
figB1, axB1 = plt.subplots(figsize=(12,5.6))
axB1.plot(fut.index, fut["DE_fut"], label="DE futures")
axB1.plot(fut.index, fut["US_fut"], label="US futures")
de_entries_B = panelB.index[long_DE_B]; us_entries_B = panelB.index[long_US_B]
axB1.scatter(de_entries_B, fut.loc[de_entries_B, "DE_fut"], marker="^", s=90, label="ENTRY: Long DE (|Z_yield|≥1)", zorder=5)
axB1.scatter(us_entries_B, fut.loc[us_entries_B, "US_fut"], marker="v", s=90, label="ENTRY: Long US (|Z_yield|≥1)", zorder=5)
axB1.set_title("Strategy B Entries — Yield-only z rule (±1σ)")
axB1.set_ylabel("Price"); axB1.set_xlabel("Date"); axB1.grid(True, alpha=0.3)
axB1.xaxis.set_major_formatter(DATE_FMT_M)
annotate_last(axB1, fut.index, fut["DE_fut"].values, fmt="{:.2f}")
annotate_last(axB1, fut.index, fut["US_fut"].values, fmt="{:.2f}")
axB1.legend(); finalize(figB1)

# Equity, returns, drawdown for B
figB2, axB2 = plt.subplots(figsize=(11,5.0))
axB2.plot(panelB.index, equityB, label="Equity")
axB2.set_title("Strategy B — Equity Curve (Yield-only ±1σ)")
axB2.set_ylabel("Growth of $1"); axB2.set_xlabel("Date"); axB2.grid(True, alpha=0.3)
axB2.xaxis.set_major_formatter(DATE_FMT_M)
annotate_last(axB2, panelB.index, equityB.values, fmt="{:.3f}")
axB2.legend(); finalize(figB2)

figB3, axB3 = plt.subplots(figsize=(11,4.2))
colorsB = np.where(panelB["Strat_ret"] >= 0, "tab:blue", "tab:orange")
axB3.bar(panelB.index, panelB["Strat_ret"].values, width=20, align="center", color=colorsB)
axB3.axhline(0, linewidth=1)
axB3.set_title("Strategy B — Monthly Returns (Wins blue / Losses orange)")
axB3.set_ylabel("Return"); axB3.set_xlabel("Date"); axB3.grid(True, alpha=0.25)
axB3.xaxis.set_major_formatter(DATE_FMT_M)
finalize(figB3)

figB4, axB4 = plt.subplots(figsize=(11,3.8))
axB4.plot(panelB.index, drawdownB)
axB4.set_title("Strategy B — Drawdown")
axB4.set_ylabel("Drawdown"); axB4.set_xlabel("Date"); axB4.grid(True, alpha=0.3)
axB4.xaxis.set_major_formatter(DATE_FMT_M)
finalize(figB4)

# ---------------- Per-trade P&L tables + Metrics ----------------
tradesA_out = tradesA[["Leg","DE_ret_fwd","US_ret_fwd","Strat_ret"]].copy()
tradesA_out.index.name = "Date"
print("\n=== Strategy A — Per-Trade P&L (single-signal futures z) ===")
print(tradesA_out.to_string(float_format=lambda x: f"{x:,.4f}"))

tradesB_out = tradesB[["Leg","Z_yield","DE_ret_fwd","US_ret_fwd","Strat_ret"]].copy()
tradesB_out.index.name = "Date"
print("\n=== Strategy B — Per-Trade P&L (yield-only ±1σ) ===")
print(tradesB_out.to_string(float_format=lambda x: f"{x:,.4f}"))

print("\n=== Strategy A — Metrics ===")
print(f"Futures z source: {FUT_SIGNAL_SOURCE.upper()}  |  Months: {len(panelA)}  |  Active months: {len(tradesA_out)}")
print(f"Cumulative Gain: {metricsA['cum_gain']:.2%}")
print(f"Max Drawdown: {metricsA['max_dd'] if 'max_dd' in metricsA else (equityA/equityA.cummax()-1).min():.2%}")
print(f"Win/Loss ratio: {metricsA['win_loss_ratio']:.3f}  (Wins={metricsA['wins']}, Losses={metricsA['losses']})")
print(f"VaR 95% (1-month): {metricsA['var95']:.2%}")
print(f"Annualized Volatility: {metricsA['ann_vol']:.2%}")
print(f"CAGR: {metricsA['cagr']:.2%}")
print(f"Total Risk Ratio (CAGR / Ann.Vol): {metricsA['trr']:.3f}")

print("\n=== Strategy B — Metrics ===")
print(f"Months: {len(panelB)}  |  Active months: {len(tradesB_out)}")
print(f"Cumulative Gain: {metricsB['cum_gain']:.2%}")
print(f"Max Drawdown: {metricsB['max_dd'] if 'max_dd' in metricsB else (equityB/equityB.cummax()-1).min():.2%}")
print(f"Win/Loss ratio: {metricsB['win_loss_ratio']:.3f}  (Wins={metricsB['wins']}, Losses={metricsB['losses']})")
print(f"VaR 95% (1-month): {metricsB['var95']:.2%}")
print(f"Annualized Volatility: {metricsB['ann_vol']:.2%}")
print(f"CAGR: {metricsB['cagr']:.2%}")
print(f"Total Risk Ratio (CAGR / Ann.Vol): {metricsB['trr']:.3f}")

# ---------------- Save outputs ----------------
outA = pd.concat(
    [
        yld[["Spread_bps","Z_yield"]],
        fut[["Spread_1x1","Z_1x1","Spread_hedged","Z_hedged"]],
        panelA[["DE_ret_fwd","US_ret_fwd","Strat_ret"]],
        equityA.rename("Equity_A"),
        drawdownA.rename("Drawdown_A"),
    ], axis=1
)
outB = pd.concat(
    [
        yld[["Spread_bps","Z_yield"]],
        fut[["DE_fut","US_fut"]],
        panelB[["Z_yield","DE_ret_fwd","US_ret_fwd","Strat_ret"]],
        equityB.rename("Equity_B"),
        drawdownB.rename("Drawdown_B"),
    ], axis=1
)
outA.to_csv("/content/strategyA_single_signal_output.csv")
outB.to_csv("/content/strategyB_yield1sigma_output.csv")

# Full OLS summaries (yield & futures)
print("\n=== Yield OLS Summary ==="); print(ols_yield.summary())
print("\n=== Futures OLS (hedge) Summary ==="); print(ols_fut.summary())

```
