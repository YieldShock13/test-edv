```python
# ============================================================
# Credit Spreads & Yields  Full Chart Suite (2000present)
# Adds HY OAS  IG OAS spread
# ============================================================

import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# Style / Theme
# -----------------------------
OCEAN_BG   = "#0b1e2d"
C_HY       = "#17becf"  # HY lines
C_IG       = "#ff7f0e"  # IG lines
C_T10      = "#2ca02c"  # 10Y Treasury
C_WHITE    = "white"    # Fed Funds (forced white)
WATERMARK  = "Orca's Capital Strategies"

def ocean_axes(ax, title):
    ax.set_facecolor(OCEAN_BG)
    ax.figure.set_facecolor(OCEAN_BG)
    ax.set_title(title, color="white", fontsize=14, fontweight="bold", pad=14)
    ax.tick_params(colors="white")
    for side in ("bottom","left"):
        ax.spines[side].set_color("white")
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.grid(alpha=0.20, color="white", linewidth=0.6)

def add_watermark(ax):
    ax.text(0.5, 0.05, WATERMARK, transform=ax.transAxes,
            fontsize=14, color="white", alpha=0.15,
            ha="center", va="center", rotation=30)

def plot_line(ax, s, label, color=None, lw=1.8, z=3):
    s = s.dropna()
    ax.plot(s.index, s.values, label=label, color=color, linewidth=lw, zorder=z)
    if not s.empty:
        x, y = s.index[-1], s.iloc[-1]
        ax.scatter([x],[y], color=color if color else "red", s=28, zorder=z+1)
        ax.annotate(f"{y:.2f}", xy=(x,y), xytext=(8,0), textcoords="offset points",
                    color="white", fontsize=9, weight="bold", zorder=z+2)

# -----------------------------
# Robust CSV Loader
# -----------------------------
def load_series_csv(path, monthly=True):
    df = pd.read_csv(path)
    date_col = df.columns[0]
    value_cols = [c for c in df.columns if c != date_col]
    val_col = value_cols[0]
    df[date_col] = pd.to_datetime(df[date_col], errors="coerce")
    df[val_col]  = pd.to_numeric(df[val_col], errors="coerce")
    s = df.set_index(date_col)[val_col].sort_index().dropna()
    if monthly:
        s = s.resample("ME").last()
    return s

# -----------------------------
# Load Data
# -----------------------------
fedfunds = load_series_csv("FEDFUNDS (2).csv")
hy_oas   = load_series_csv("BAMLH0A0HYM2 (10).csv")
ig_oas   = load_series_csv("BAMLC0A0CM (3).csv")
dgs10    = load_series_csv("DGS10 (12).csv")
hy_ey    = load_series_csv("BAMLH0A0HYM2EY.csv")
ig_ey    = load_series_csv("BAMLC0A0CMEY (1).csv")

data = pd.concat(
    [fedfunds.rename("FedFunds"), hy_oas.rename("HY_OAS"), ig_oas.rename("IG_OAS"),
     dgs10.rename("DGS10"), hy_ey.rename("HY_EY"), ig_ey.rename("IG_EY")],
    axis=1
).sort_index()
data = data.loc["2000":]

# -----------------------------
# Derived Spreads
# -----------------------------
data["IG_minus_Treas"] = data["IG_EY"] - data["DGS10"]
data["HY_minus_IG"]    = data["HY_EY"] - data["IG_EY"]
data["HY_minus_Treas"] = data["HY_EY"] - data["DGS10"]
data["HY_OAS_minus_IG_OAS"] = data["HY_OAS"] - data["IG_OAS"]  # NEW

# -----------------------------
# Charts
# -----------------------------
# Chart 1
fig, ax = plt.subplots(figsize=(12,6))
plot_line(ax, data["HY_EY"],  "HY Effective Yield", color=C_HY)
plot_line(ax, data["IG_EY"],  "IG Effective Yield", color=C_IG)
plot_line(ax, data["DGS10"],  "10Y Treasury",       color=C_T10)
ocean_axes(ax, "Effective Yields vs 10Y Treasury (2000)")
ax.legend(facecolor=OCEAN_BG, edgecolor="none", labelcolor="white")
add_watermark(ax); plt.tight_layout(); plt.show()

# Chart 2
fig, ax = plt.subplots(figsize=(12,6))
plot_line(ax, data["HY_OAS"], "HY OAS", color=C_HY)
plot_line(ax, data["IG_OAS"], "IG OAS", color=C_IG)
ocean_axes(ax, "Option-Adjusted Spreads (2000)")
ax.legend(facecolor=OCEAN_BG, edgecolor="none", labelcolor="white")
add_watermark(ax); plt.tight_layout(); plt.show()

# Chart 3
fig, ax = plt.subplots(figsize=(12,6))
plot_line(ax, data["IG_minus_Treas"], "IG Yield  10Y Treasury", color=C_IG)
ocean_axes(ax, "Spread: IG Effective Yield  10Y Treasury (2000)")
add_watermark(ax); plt.tight_layout(); plt.show()

# Chart 4
fig, ax = plt.subplots(figsize=(12,6))
plot_line(ax, data["HY_minus_IG"], "HY Effective Yield  IG Effective Yield", color=C_HY)
ocean_axes(ax, "Spread: HY Effective Yield  IG Effective Yield (2000)")
add_watermark(ax); plt.tight_layout(); plt.show()

# Chart 5
fig, ax = plt.subplots(figsize=(12,6))
plot_line(ax, data["HY_minus_Treas"], "HY Effective Yield  10Y Treasury", color=C_HY)
ocean_axes(ax, "Spread: HY Effective Yield  10Y Treasury (2000)")
add_watermark(ax); plt.tight_layout(); plt.show()

# Chart 6
fig, ax = plt.subplots(figsize=(12,6))
plot_line(ax, data["HY_OAS"], "HY OAS", color=C_HY, lw=1.9, z=5)
plot_line(ax, data["FedFunds"], "Fed Funds Rate", color=C_WHITE, lw=2.4, z=10)
ocean_axes(ax, "Fed Funds (WHITE) vs HY OAS (2000)")
ax.legend(facecolor=OCEAN_BG, edgecolor="none", labelcolor="white")
add_watermark(ax); plt.tight_layout(); plt.show()

# Chart 7 (NEW)
fig, ax = plt.subplots(figsize=(12,6))
plot_line(ax, data["HY_OAS_minus_IG_OAS"], "HY OAS  IG OAS", color="#d62728")
ocean_axes(ax, "Spread: HY OAS  IG OAS (2000)")
add_watermark(ax); plt.tight_layout(); plt.show()





```
