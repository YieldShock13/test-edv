```python
# ============================================================
# Pearl Gray Equity and Research
# US Rates (weekly): 2Y / 10Y / 20Y
# 52-week rolling framework + Bloomberg-style dark charts
#
# IMPORTANT FIX:
# - Weekly resample is labeled by the ACTUAL last observation date in each week,
#   NOT the week-ending Friday label (prevents "future" dates like 2026-01-30).
#
# Butterfly (as specified):
#   +0.5 * 2Y  +0.5 * 20Y  -1.0 * 10Y
#
# Rolling spread charts (52W rolling mean):
#   20Y - 2Y
#   10Y - 2Y
#   20Y - 10Y
#
# Z-score / Vol (52W rolling) computed on RAW spreads
#
# Half-life (your definition, event-based):
#   mu_t = 52W moving average of spread, d_t = s_t - mu_t.
#   half-life at t = smallest h such that:
#       |d_{t+h}| <= 0.5*|d_t|
#   search forward up to MAX_H weeks, else blank (NaN).
# ============================================================

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# ---------------------------
# CONFIG
# ---------------------------
WATERMARK   = "Orca's Capital Strategies"
ROLL_W      = 52
WEEKLY_RULE = "W-FRI"   # grouping frequency (week ending Friday), but we will relabel by actual last obs date
ANNUALIZE_VOL = True
MAX_H       = 104       # event-based half-life search horizon (weeks); if not hit => NaN (blank)

FILE_2Y  = "United States 2-Year Bond Yield Historical Data (3).csv"
FILE_10Y = "United States 10-Year Bond Yield Historical Data (12).csv"
FILE_20Y = "U.S. 20-Year Bond Yield Historical Data.csv"

SEARCH_ROOTS = [".", "/mnt/data", "/content"]

# ---------------------------
# BLOOMBERG-STYLE DARK THEME
# ---------------------------
plt.rcParams.update({
    "figure.dpi": 140,
    "savefig.dpi": 140,

    "figure.facecolor": "black",
    "axes.facecolor": "black",
    "savefig.facecolor": "black",

    "text.color": "white",
    "axes.labelcolor": "white",
    "axes.edgecolor": "white",
    "axes.titlecolor": "white",
    "xtick.color": "white",
    "ytick.color": "white",

    "font.family": "DejaVu Sans Mono",
    "axes.grid": True,
    "grid.color": "#3a3a3a",
    "grid.alpha": 0.35,

    "axes.titlesize": 12,
    "axes.labelsize": 10,
    "xtick.labelsize": 9,
    "ytick.labelsize": 9,
    "lines.linewidth": 1.8,
})

# ---------------------------
# IO HELPERS
# ---------------------------
def _find_path(filename: str) -> str:
    for root in SEARCH_ROOTS:
        p = os.path.join(root, filename)
        if os.path.exists(p):
            return p
    raise FileNotFoundError(f"Could not find '{filename}' in {SEARCH_ROOTS}")

def _to_float_series(s: pd.Series) -> pd.Series:
    s = s.astype(str).str.replace(",", "", regex=False).str.replace("%", "", regex=False).str.strip()
    s = s.replace({"": np.nan, "nan": np.nan, "NaN": np.nan, "-": np.nan})
    return pd.to_numeric(s, errors="coerce")

def load_investing_yield_csv(filename: str) -> pd.Series:
    path = _find_path(filename)
    df = pd.read_csv(path)

    # Date column
    date_col = None
    for c in df.columns:
        if c.strip().lower() == "date":
            date_col = c
            break
    if date_col is None:
        raise ValueError(f"{filename}: missing 'Date' column. Columns: {list(df.columns)}")

    # Value column (Investing.com yield pages often store yield level under "Price")
    y_col = None
    for cand in ["Price", "Last", "Close", "Yield", "Rate"]:
        if cand in df.columns:
            y_col = cand
            break
    if y_col is None:
        # fallback: first non-date column
        for c in df.columns:
            if c != date_col:
                y_col = c
                break

    # parse dates robustly
    d1 = pd.to_datetime(df[date_col], errors="coerce", dayfirst=True)
    if d1.notna().mean() < 0.5:
        d1 = pd.to_datetime(df[date_col], errors="coerce", dayfirst=False)

    df = df.assign(_date=d1).dropna(subset=["_date"]).sort_values("_date")

    y = _to_float_series(df[y_col])
    s = pd.Series(y.values, index=pd.DatetimeIndex(df["_date"]), name=os.path.splitext(filename)[0]).dropna()
    return s

def to_weekly_last_actual_date(s: pd.Series, rule: str = WEEKLY_RULE) -> pd.Series:
    """
    Groups observations into weekly buckets (week ending Friday), takes the last obs in each bucket,
    then *relabels the index with the actual date of that last obs* (prevents 'future' week-end labels).
    """
    s = s.sort_index().dropna()
    g = s.groupby(pd.Grouper(freq=rule))

    last_vals = g.apply(lambda x: x.iloc[-1] if len(x) else np.nan).dropna()
    last_dates = g.apply(lambda x: x.index[-1] if len(x) else pd.NaT).dropna()

    last_vals.index = pd.DatetimeIndex(last_dates.values)
    return last_vals.sort_index()

# ---------------------------
# PLOTTING HELPERS
# ---------------------------
def _watermark(ax, text=WATERMARK):
    ax.text(
        0.5, 0.5, text,
        transform=ax.transAxes,
        fontsize=28,
        color="#bfbfbf",
        alpha=0.10,
        ha="center",
        va="center",
        rotation=25
    )

def _label_last_point(ax, x: pd.Series, fmt="{:.3f}"):
    x = x.dropna()
    if x.empty:
        return
    last_idx = x.index[-1]
    last_val = float(x.iloc[-1])
    ax.scatter([last_idx], [last_val], s=28)
    ax.annotate(
        f"{last_idx.date()}  {fmt.format(last_val)}",
        xy=(last_idx, last_val),
        xytext=(10, 0),
        textcoords="offset points",
        ha="left",
        va="center",
        fontsize=9,
        color="white",
        bbox=dict(boxstyle="round,pad=0.25", fc="#111111", ec="#666666", alpha=0.85),
    )

def plot_series(x: pd.Series, title: str, ylabel: str, ylim=None):
    fig, ax = plt.subplots(figsize=(12.6, 4.6))
    ax.plot(x.index, x.values)

    ax.set_title(title)
    ax.set_ylabel(ylabel)

    ax.xaxis.set_major_locator(mdates.YearLocator(2))
    ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))

    for spine in ax.spines.values():
        spine.set_color("white")

    if ylim is not None:
        ax.set_ylim(ylim)

    _watermark(ax)
    _label_last_point(ax, x)

    plt.tight_layout()
    plt.show()

# ---------------------------
# ROLLING STATS
# ---------------------------
def rolling_mean(x: pd.Series, w: int = ROLL_W) -> pd.Series:
    return x.rolling(w).mean()

def rolling_std(x: pd.Series, w: int = ROLL_W) -> pd.Series:
    return x.rolling(w).std(ddof=0)

def rolling_z(x: pd.Series, w: int = ROLL_W) -> pd.Series:
    mu = rolling_mean(x, w)
    sd = rolling_std(x, w)
    return (x - mu) / sd

def rolling_vol(x: pd.Series, w: int = ROLL_W, annualize: bool = ANNUALIZE_VOL) -> pd.Series:
    v = rolling_std(x, w)
    return v * np.sqrt(52) if annualize else v

# ---------------------------
# HALF-LIFE (your definition, event-based)
# ---------------------------
def event_half_life_from_ma(spread: pd.Series, w: int = ROLL_W, max_h: int = MAX_H) -> pd.Series:
    """
    mu_t = rolling mean of spread over w weeks
    d_t  = spread_t - mu_t
    half-life at t = smallest h>=1 such that:
        |d_{t+h}| <= 0.5*|d_t|
    Search forward up to max_h observations; else NaN.
    """
    mu = rolling_mean(spread, w)
    d  = (spread - mu)

    idx = spread.index
    dvals = d.values
    hl = np.full(len(spread), np.nan, dtype=float)

    for i in range(len(spread)):
        if np.isnan(dvals[i]):
            continue

        d0 = abs(dvals[i])
        if d0 == 0:
            hl[i] = 0.0
            continue

        target = 0.5 * d0
        j_end = min(len(spread) - 1, i + max_h)

        for j in range(i + 1, j_end + 1):
            if np.isnan(dvals[j]):
                continue
            if abs(dvals[j]) <= target:
                hl[i] = float(j - i)
                break

    return pd.Series(hl, index=idx, name=f"Half-life to halve |spread - MA{w}| (weeks)")

# ============================================================
# LOAD DATA (weekly indexed by ACTUAL last obs date)
# ============================================================
y2  = to_weekly_last_actual_date(load_investing_yield_csv(FILE_2Y)).rename("US2Y")
y10 = to_weekly_last_actual_date(load_investing_yield_csv(FILE_10Y)).rename("US10Y")
y20 = to_weekly_last_actual_date(load_investing_yield_csv(FILE_20Y)).rename("US20Y")

df = pd.concat([y2, y10, y20], axis=1).dropna().sort_index()

data_start = df.index.min().date()
data_end   = df.index.max().date()
print(f"Weekly (actual-date) data range used: {data_start} to {data_end}")
print(f"Weekly observations: {len(df):,}")

# ============================================================
# CONSTRUCT SERIES
# ============================================================
butterfly = (0.5*df["US2Y"] + 0.5*df["US20Y"] - 1.0*df["US10Y"]).rename(
    "Butterfly: 0.5*2Y + 0.5*20Y - 1.0*10Y"
)

spr_20_2  = (df["US20Y"] - df["US2Y"]).rename("Spread: 20Y - 2Y")
spr_10_2  = (df["US10Y"] - df["US2Y"]).rename("Spread: 10Y - 2Y")
spr_20_10 = (df["US20Y"] - df["US10Y"]).rename("Spread: 20Y - 10Y")

# Rolling spread charts (52W rolling mean)  your requirement that spreads be rolling
spr_20_2_roll  = rolling_mean(spr_20_2, ROLL_W).rename("ROLLING MEAN(52W): 20Y - 2Y")
spr_10_2_roll  = rolling_mean(spr_10_2, ROLL_W).rename("ROLLING MEAN(52W): 10Y - 2Y")
spr_20_10_roll = rolling_mean(spr_20_10, ROLL_W).rename("ROLLING MEAN(52W): 20Y - 10Y")

# Z / Vol (on RAW spreads)
z_20_2   = rolling_z(spr_20_2, ROLL_W).rename("Z(52W): 20Y - 2Y")
z_10_2   = rolling_z(spr_10_2, ROLL_W).rename("Z(52W): 10Y - 2Y")
z_20_10  = rolling_z(spr_20_10, ROLL_W).rename("Z(52W): 20Y - 10Y")

vol_20_2  = rolling_vol(spr_20_2, ROLL_W, ANNUALIZE_VOL).rename("VOL(52W): 20Y - 2Y")
vol_10_2  = rolling_vol(spr_10_2, ROLL_W, ANNUALIZE_VOL).rename("VOL(52W): 10Y - 2Y")
vol_20_10 = rolling_vol(spr_20_10, ROLL_W, ANNUALIZE_VOL).rename("VOL(52W): 20Y - 10Y")

# Half-life (event-based, your method)
hl_20_2  = event_half_life_from_ma(spr_20_2,  ROLL_W, MAX_H).rename("Half-life: 20Y - 2Y")
hl_10_2  = event_half_life_from_ma(spr_10_2,  ROLL_W, MAX_H).rename("Half-life: 10Y - 2Y")
hl_20_10 = event_half_life_from_ma(spr_20_10, ROLL_W, MAX_H).rename("Half-life: 20Y - 10Y")

# ============================================================
# PLOTS (ALL SEPARATE CHARTS)
# ============================================================

# Butterfly
plot_series(butterfly, f"{butterfly.name} (Weekly) | Data: {data_start} to {data_end}", "pct points")

# Rolling spread charts (separate)
plot_series(spr_20_2_roll,  f"{spr_20_2_roll.name} | {ROLL_W}W | Data: {data_start} to {data_end}", "pct points")
plot_series(spr_10_2_roll,  f"{spr_10_2_roll.name} | {ROLL_W}W | Data: {data_start} to {data_end}", "pct points")
plot_series(spr_20_10_roll, f"{spr_20_10_roll.name} | {ROLL_W}W | Data: {data_start} to {data_end}", "pct points")

# Z-score charts (separate)
plot_series(z_20_2,  f"{z_20_2.name} | {ROLL_W}W | Data: {data_start} to {data_end}", "z")
plot_series(z_10_2,  f"{z_10_2.name} | {ROLL_W}W | Data: {data_start} to {data_end}", "z")
plot_series(z_20_10, f"{z_20_10.name} | {ROLL_W}W | Data: {data_start} to {data_end}", "z")

# Vol charts (separate)
vol_label = "pct points (annualized)" if ANNUALIZE_VOL else "pct points (weekly std)"
plot_series(vol_20_2,  f"{vol_20_2.name} | {ROLL_W}W | Data: {data_start} to {data_end}", vol_label)
plot_series(vol_10_2,  f"{vol_10_2.name} | {ROLL_W}W | Data: {data_start} to {data_end}", vol_label)
plot_series(vol_20_10, f"{vol_20_10.name} | {ROLL_W}W | Data: {data_start} to {data_end}", vol_label)

# Half-life charts (separate)
plot_series(hl_20_2,  f"{hl_20_2.name} | time to halve |spread - MA52| | max_h={MAX_H}", "weeks")
plot_series(hl_10_2,  f"{hl_10_2.name} | time to halve |spread - MA52| | max_h={MAX_H}", "weeks")
plot_series(hl_20_10, f"{hl_20_10.name} | time to halve |spread - MA52| | max_h={MAX_H}", "weeks")




```
