```python
# ======================================================
# ORCA  VAR IRFs ONLY (Orthogonalized) with 99%, 75%, 50% bands
# OVX (Oil Volatility) / VIX / MOVE
# Notes:
#  - Y-axis = impulse response (units per 1-unit shock)  time-varying beta.
#  - X-axis = Days after shock (h), labeled at EVERY day: 0..H.
# ======================================================

import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.api import VAR

# ----------------------------
# 1) LOAD & ALIGN THE DATA
# ----------------------------
OVX_FILE  = "OVXCLS (2).csv"
VIX_FILE  = "VIXCLS (2).csv"
MOVE_FILE = "ICE BofAML MOVE Historical Data.csv"

def load_fred(path, newname):
    df = pd.read_csv(path)
    df["Date"] = pd.to_datetime(df["observation_date"])
    df[newname] = pd.to_numeric(df.iloc[:, 1], errors="coerce")
    return df[["Date", newname]].set_index("Date").sort_index()

def load_move(path):
    df = pd.read_csv(path)
    if "Date" not in df.columns:
        dcol = [c for c in df.columns if "date" in c.lower()][0]
        df.rename(columns={dcol: "Date"}, inplace=True)
    df["Date"] = pd.to_datetime(df["Date"])
    # Choose the most-complete numeric column
    best, best_n = None, -1
    for c in df.columns:
        if c == "Date": continue
        s = pd.to_numeric(df[c].astype(str).str.replace(",",""), errors="coerce")
        if s.notna().sum() > best_n:
            best, best_n = c, s.notna().sum()
            df[c] = s
    return df.set_index("Date")[[best]].rename(columns={best: "MOVE Index"}).sort_index()

ovx  = load_fred(OVX_FILE, "Oil Volatility")
vix  = load_fred(VIX_FILE, "VIX")
move = load_move(MOVE_FILE)

# Align to business days and forward-fill
lvl = pd.concat([ovx, vix, move], axis=1).asfreq("B").ffill()

# Use daily levels (more stationary for VAR)
rets = lvl.diff().dropna()

SERIES = ["Oil Volatility", "VIX", "MOVE Index"]
H = 15  # horizon in days (0..H shown)

# ----------------------------
# 2) FIT VAR (lag by AIC, capped)
# ----------------------------
model = VAR(rets[SERIES].dropna())
sel = model.select_order(20)
try:
    p_opt = int(sel.aic)
except Exception:
    p_opt = int(getattr(sel, "selected_orders", {}).get("aic", 5)) if hasattr(sel, "selected_orders") else 5
p_opt = max(1, min(p_opt, 20))
print(f"Optimal lag order (AIC): {p_opt}")

var_res = model.fit(p_opt)
irf = var_res.irf(H)  # IR analysis object (orth_irfs available)

# ----------------------------
# 3) PLOTTING: 33 IRF grid for a given confidence level
# ----------------------------
def plot_irf_grid(irf_obj, series_names, signif=0.01, title_prefix="Impulse responses (orthogonalized)"):
    """
    signif: 0.01 = 99% bands, 0.25 = 75%, 0.50 = 50%
    """
    n = len(series_names)
    low, high = irf_obj.errband_mc(orth=True, repl=500, signif=signif)  # shapes: (steps+1, n, n)
    steps = np.arange(0, H+1)

    fig, axes = plt.subplots(n, n, figsize=(12, 12))
    fig.suptitle(f"{title_prefix}  CI={(1 - signif):.0%}", fontsize=14, y=0.94)

    for i_resp, resp in enumerate(series_names):
        for j_shock, shock in enumerate(series_names):
            ax = axes[i_resp, j_shock]
            vals = irf_obj.orth_irfs[:H+1, i_resp, j_shock]
            lo = low [:H+1, i_resp, j_shock]
            hi = high[:H+1, i_resp, j_shock]

            ax.plot(steps, vals, color="C0", lw=2)
            ax.fill_between(steps, lo, hi, color="C0", alpha=0.2, label=f"{int((1-signif)*100)}% CI")
            ax.axhline(0, color="k", ls="--", lw=1)

            ax.set_title(f"{shock}  {resp}", fontsize=10)
            # X axis: label EVERY day 0..H
            ax.set_xlim(0, H)
            ax.set_xticks(np.arange(0, H+1, 1))
            ax.tick_params(axis="x", labelrotation=0, labelsize=8)
            ax.set_xlabel("Days after shock (h)", fontsize=9)

            # Y axis label only on leftmost column for cleanliness
            if j_shock == 0:
                ax.set_ylabel("IRF (units per 1-unit shock)", fontsize=9)

    handles, labels = axes[0,0].get_legend_handles_labels()
    if handles:
        fig.legend(handles, labels, loc="lower center", ncol=3, bbox_to_anchor=(0.5, 0.02))
    fig.tight_layout(rect=[0, 0.04, 1, 0.92])
    plt.show()

# ----------------------------
# 4) PLOT at 99%, 75%, and 50% CONFIDENCE
# ----------------------------
plot_irf_grid(irf, SERIES, signif=0.01, title_prefix="Impulse responses (orthogonalized)")
plot_irf_grid(irf, SERIES, signif=0.25, title_prefix="Impulse responses (orthogonalized)")
plot_irf_grid(irf, SERIES, signif=0.50, title_prefix="Impulse responses (orthogonalized)")












```
