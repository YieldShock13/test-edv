```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# =========================
# SETTINGS
# =========================
FILE_NAME = "WTI Crude - Brent Spread c1 Historical Data (3).csv"
WINDOW = 252

# =========================
# LOAD + CLEAN
# =========================
df = pd.read_csv(FILE_NAME)
df.columns = [c.strip() for c in df.columns]

df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
df["Price"] = (
    df["Price"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .str.replace("%", "", regex=False)
)
df["Price"] = pd.to_numeric(df["Price"], errors="coerce")

df = df.dropna(subset=["Date", "Price"]).copy()
df = df.sort_values("Date").drop_duplicates(subset="Date")
df = df.set_index("Date")

# Rename for clarity
df["spread"] = df["Price"]

# =========================
# ROLLING Z-SCORE ON SPREAD LEVEL
# =========================
roll_mean = df["spread"].rolling(WINDOW, min_periods=WINDOW).mean()
roll_std = df["spread"].rolling(WINDOW, min_periods=WINDOW).std()

df["zscore_252"] = (df["spread"] - roll_mean) / roll_std

# =========================
# OPTIONAL: FINAL DATA
# =========================
last_date = df.index[-1]
last_spread = df["spread"].iloc[-1]
last_z = df["zscore_252"].dropna().iloc[-1] if df["zscore_252"].notna().any() else np.nan
last_z_date = df["zscore_252"].dropna().index[-1] if df["zscore_252"].notna().any() else None

print(f"Last spread: {last_spread:.2f} on {last_date.date()}")
if pd.notna(last_z):
    print(f"Last 252d z-score: {last_z:.2f} on {last_z_date.date()}")

# =========================
# STYLE
# =========================
plt.style.use("seaborn-v0_8-darkgrid")

def label_last_point(ax, x, y, text):
    ax.scatter(x, y, s=35, zorder=5)
    ax.annotate(
        text,
        xy=(x, y),
        xytext=(8, 0),
        textcoords="offset points",
        va="center",
        fontsize=10,
        bbox=dict(boxstyle="round,pad=0.25", fc="white", ec="gray", alpha=0.9)
    )

# =========================
# CHART 1: SPREAD LEVEL
# =========================
fig, ax = plt.subplots(figsize=(15, 7))
ax.plot(df.index, df["spread"], linewidth=2)
ax.axhline(df["spread"].mean(), linestyle="--", linewidth=1)

ax.set_title("WTI–Brent Spread", fontsize=17, pad=15)
ax.set_xlabel("")
ax.set_ylabel("Spread")
label_last_point(ax, df.index[-1], df["spread"].iloc[-1], f"{df['spread'].iloc[-1]:.2f}")

plt.tight_layout()
plt.show()

# =========================
# CHART 2: 252D ROLLING Z-SCORE
# =========================
z = df.dropna(subset=["zscore_252"]).copy()

fig, ax = plt.subplots(figsize=(15, 7))
ax.plot(z.index, z["zscore_252"], linewidth=2)
ax.axhline(0, linestyle="--", linewidth=1)
ax.axhline(1, linestyle="--", linewidth=1)
ax.axhline(-1, linestyle="--", linewidth=1)
ax.axhline(2, linestyle="--", linewidth=1)
ax.axhline(-2, linestyle="--", linewidth=1)

ax.set_title("WTI–Brent Spread: 252-Day Rolling Z-Score", fontsize=17, pad=15)
ax.set_xlabel("")
ax.set_ylabel("Z-Score")

if not z.empty:
    label_last_point(ax, z.index[-1], z["zscore_252"].iloc[-1], f"{z['zscore_252'].iloc[-1]:.2f}")

plt.tight_layout()
plt.show()

# =========================
# MEANS (FULL / 5Y / 3Y)
# =========================

# Ensure sorted index
df = df.sort_index()

# Full sample mean
mean_full = df["spread"].mean()

# Last 5 years
cut_5y = df.index.max() - pd.DateOffset(years=5)
df_5y = df[df.index >= cut_5y]
mean_5y = df_5y["spread"].mean()

# Last 3 years
cut_3y = df.index.max() - pd.DateOffset(years=3)
df_3y = df[df.index >= cut_3y]
mean_3y = df_3y["spread"].mean()

# =========================
# OUTPUT
# =========================
print("MEAN (SPREAD):")
print(f"Full sample   : {mean_full:.4f}")
print(f"Last 5 years  : {mean_5y:.4f}")
print(f"Last 3 years  : {mean_3y:.4f}")
```
