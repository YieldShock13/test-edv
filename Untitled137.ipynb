```python
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
import statsmodels.api as sm

# =========================
# 0) FILES (plain filenames)
# =========================
DATA_DIR = Path(".")  # folder where the CSVs live

FILES = {
    "USDZAR": "USD_ZAR Historical Data (3).csv",
    "VIX":    "CBOE Volatility Index Historical Data (9).csv",
    "DXY":    "US Dollar Index Historical Data (3).csv",
    "GOLD":   "Gold Futures Historical Data (6).csv",
    "PLAT":   "Platinum Futures Historical Data (2).csv",
    "COPPER": "Copper Futures Historical Data (1).csv",
    "US3M":   "United States 3-Month Bond Yield Historical Data (1).csv",
    "ZA3M":   "South Africa 3-Month Bond Yield Historical Data (2).csv",
    "ZA_CDS": "ZAGV10YUSAC=R Overview.csv",
    "US_CDS": "USGV10YUSAB=R Overview (5).csv",
}

missing = [f for f in FILES.values() if not (DATA_DIR / f).exists()]
if missing:
    raise FileNotFoundError("Missing files in DATA_DIR:\n" + "\n".join(missing))

# ===========================================
# 1) Load + clean Investing.com-style CSVs
# ===========================================
def read_investing_csv(filename: str) -> pd.Series:
    p = DATA_DIR / filename
    df = pd.read_csv(p)
    df.columns = [c.strip() for c in df.columns]

    date_col = next((c for c in df.columns if c.lower() == "date"), df.columns[0])
    price_col = next((c for c in df.columns if c.lower() in ["price", "close", "last"]), df.columns[1])

    out = df[[date_col, price_col]].copy()
    out.columns = ["Date", "Value"]
    out["Date"] = pd.to_datetime(out["Date"], errors="coerce")

    out["Value"] = (out["Value"].astype(str)
                    .str.replace(",", "", regex=False)
                    .str.replace("%", "", regex=False)
                    .str.strip())
    out["Value"] = pd.to_numeric(out["Value"], errors="coerce")

    out = (out.dropna(subset=["Date", "Value"])
              .drop_duplicates("Date")
              .sort_values("Date")
              .set_index("Date")["Value"])
    return out

S = {k: read_investing_csv(v).rename(k) for k, v in FILES.items()}

# ===========================================
# 2) Align on USDZAR weekly calendar (master)
# ===========================================
df = S["USDZAR"].to_frame()
for k, ser in S.items():
    if k == "USDZAR":
        continue
    df[k] = ser.reindex(df.index).ffill()
df = df.sort_index()

# ===========================================
# 3) Movement vs State
# ===========================================
df["fx_ret"]    = np.log(df["USDZAR"]).diff()
df["dxy_ret"]   = np.log(df["DXY"]).diff()
df["gold_ret"]  = np.log(df["GOLD"]).diff()
df["plat_ret"]  = np.log(df["PLAT"]).diff()
df["copp_ret"]  = np.log(df["COPPER"]).diff()

df["carry_bps"]      = (df["ZA3M"] - df["US3M"]) * 100.0
df["cds_spread_bps"] = (df["ZA_CDS"] - df["US_CDS"])

df["vix_lag"]   = df["VIX"].shift(1)
df["carry_lag"] = df["carry_bps"].shift(1)
df["cds_lag"]   = df["cds_spread_bps"].shift(1)

# ===========================================
# 4) Regimes from VIX level
# ===========================================
v = df["VIX"].dropna()
q30, q70 = v.quantile(0.30), v.quantile(0.70)
df["regime"] = np.where(df["VIX"] >= q70, "risk_off",
                 np.where(df["VIX"] <= q30, "risk_on", "neutral"))

# ===========================================
# 5) Summary stats
# ===========================================
print("\n=========================")
print("SUMMARY STATS (OVERALL)")
print("=========================")
summ_cols = ["USDZAR","VIX","DXY","GOLD","PLAT","COPPER","ZA3M","US3M","carry_bps","ZA_CDS","US_CDS","cds_spread_bps"]
print(df[summ_cols].describe().T[["count","mean","std","min","25%","50%","75%","max"]])

print("\n=========================")
print("SUMMARY STATS (RETURNS, %)")
print("=========================")
ret_cols = ["fx_ret","dxy_ret","gold_ret","plat_ret","copp_ret"]
print((100*df[ret_cols]).describe().T[["count","mean","std","min","25%","50%","75%","max"]].rename(columns={
    "mean":"mean_%","std":"std_%","min":"min_%","25%":"p25_%","50%":"median_%","75%":"p75_%","max":"max_%"
}))

print("\n=========================")
print("REGIME BEHAVIOUR (USDZAR weekly log returns)")
print("=========================")
reg_stats = (df.dropna(subset=["fx_ret"])
             .groupby("regime")
             .agg(n_weeks=("fx_ret","size"),
                  mean_fx_ret_pct=("fx_ret", lambda x: 100*x.mean()),
                  vol_fx_ret_pct=("fx_ret",  lambda x: 100*x.std()),
                  pct_usdzar_up=("fx_ret",    lambda x: 100*(x>0).mean()),
                  mean_vix=("VIX","mean"),
                  mean_cds_spread=("cds_spread_bps","mean"),
                  mean_carry_bps=("carry_bps","mean")))
print(reg_stats)

# ===========================================
# 6) Model 2 (interaction) + rolling (52w, ex-ante)
# ===========================================
d2 = df.dropna(subset=["fx_ret","dxy_ret","gold_ret","plat_ret","copp_ret","carry_lag","cds_lag","VIX","regime"]).copy()
d2["risk_off_flag"] = (d2["regime"]=="risk_off").astype(int)
d2["dxy_x_ro"] = d2["dxy_ret"] * d2["risk_off_flag"]
d2["cds_x_ro"] = d2["cds_lag"] * d2["risk_off_flag"]

Xcols = ["dxy_ret","dxy_x_ro","gold_ret","plat_ret","copp_ret","carry_lag","cds_lag","cds_x_ro"]
X = sm.add_constant(d2[Xcols])
m2 = sm.OLS(d2["fx_ret"], X).fit(cov_type="HAC", cov_kwds={"maxlags": 4})

print("\n=========================")
print("MODEL 2: INTERACTION (full-sample)")
print("=========================")
print(m2.params)
print("t-stats:")
print(m2.tvalues)

ROLL_WIN = 52
MIN_FRAC = 0.85
min_obs = int(np.ceil(ROLL_WIN * MIN_FRAC))

roll_df = d2[["fx_ret"] + Xcols].copy()

betas = []
beta_idx = []
for i in range(ROLL_WIN, len(roll_df)):
    w = roll_df.iloc[i-ROLL_WIN:i].dropna()
    if len(w) < min_obs:
        betas.append([np.nan] * (1 + len(Xcols)))
        beta_idx.append(roll_df.index[i])
        continue
    y = w["fx_ret"].values
    Xw = np.column_stack([np.ones(len(w))] + [w[c].values for c in Xcols])
    b, *_ = np.linalg.lstsq(Xw, y, rcond=None)
    betas.append(b.tolist())
    beta_idx.append(roll_df.index[i])

beta_cols = ["const"] + Xcols
beta_ts = pd.DataFrame(betas, index=pd.Index(beta_idx, name="Date"), columns=beta_cols)

apply = d2.loc[beta_ts.index].copy()

roll_contrib = pd.DataFrame(index=beta_ts.index)
for c in Xcols:
    roll_contrib[c] = apply[c].values * beta_ts[c].values
roll_contrib["const"]  = beta_ts["const"].values
roll_contrib["pred"]   = roll_contrib[Xcols].sum(axis=1) + roll_contrib["const"]
roll_contrib["actual"] = apply["fx_ret"]
roll_contrib["resid"]  = roll_contrib["actual"] - roll_contrib["pred"]

tape = pd.DataFrame(index=roll_contrib.index)
tape["DXY"]     = roll_contrib["dxy_ret"] + roll_contrib["dxy_x_ro"]
tape["Metals"]  = roll_contrib["gold_ret"] + roll_contrib["plat_ret"]
tape["Copper"]  = roll_contrib["copp_ret"]
tape["SA_Risk"] = roll_contrib["cds_lag"] + roll_contrib["cds_x_ro"]
tape["Carry"]   = roll_contrib["carry_lag"]
tape["Const"]   = roll_contrib["const"]
tape["Pred"]    = roll_contrib["pred"]
tape["Actual"]  = roll_contrib["actual"]
tape["Resid"]   = roll_contrib["resid"]

SMOOTH_W = 13
tape_s = tape.rolling(SMOOTH_W).sum()

# ===========================================
# 7) CHART STYLE + NO-OVERLAP HEADER SYSTEM
# ===========================================
plt.rcParams.update({
    "figure.facecolor": "white",
    "axes.facecolor": "white",
    "axes.edgecolor": "#444444",
    "axes.labelcolor": "#111111",
    "xtick.color": "#111111",
    "ytick.color": "#111111",
    "grid.color": "#DDDDDD",
    "text.color": "#111111",
    "font.family": "DejaVu Sans Mono",
})

REG_COL = {
    "risk_on":  (0.20, 0.65, 0.35, 0.12),
    "neutral":  (0.60, 0.60, 0.60, 0.06),
    "risk_off": (0.85, 0.25, 0.25, 0.12),
}

def fig_ax(figsize):
    fig, ax = plt.subplots(figsize=figsize, constrained_layout=False)
    # Reserve a header band so NOTHING overlaps
    fig.subplots_adjust(top=0.84, left=0.07, right=0.98, bottom=0.12)
    return fig, ax

def add_header(fig, ax, title, subtitle=None, legend_ncol=2):
    # Title + subtitle live in the FIGURE header band (not inside axes)
    fig.text(0.01, 0.975, title, ha="left", va="top", fontsize=12, fontweight="bold", color="#111111")
    if subtitle:
        fig.text(0.01, 0.94, subtitle, ha="left", va="top", fontsize=9, color="#444444")
    # Legend in header band (right) so it never collides with title/subtitle
    handles, labels = ax.get_legend_handles_labels()
    if handles:
        fig.legend(handles, labels, loc="upper right", bbox_to_anchor=(0.99, 0.985),
                   ncol=legend_ncol, frameon=True, facecolor="white", edgecolor="#BBBBBB",
                   fontsize=9)

def add_regime_shading(ax, idx, regime_series, y_min, y_max):
    # Shade contiguous runs (cleaner than per-point stripes)
    rs = pd.Series(regime_series, index=idx).dropna()
    if rs.empty:
        return
    runs = []
    cur_reg = rs.iloc[0]
    start = rs.index[0]
    for t, reg in rs.iloc[1:].items():
        if reg != cur_reg:
            runs.append((start, t, cur_reg))
            start = t
            cur_reg = reg
    runs.append((start, rs.index[-1], cur_reg))
    for a, b, reg in runs:
        rgba = REG_COL.get(reg, (0.6, 0.6, 0.6, 0.05))
        ax.axvspan(a, b, ymin=0, ymax=1, color=rgba[:3], alpha=rgba[3], lw=0)

def pad_right(ax, idx, pad_frac=0.03):
    x = pd.Index(idx)
    if len(x) < 2:
        return
    span = x[-1] - x[0]
    ax.set_xlim(x[0], x[-1] + span * pad_frac)

def label_last(ax, idx, y, fmt="{:.2f}", dy=0):
    s = pd.Series(y, index=idx).dropna()
    if s.empty:
        return
    ax.annotate(fmt.format(s.iloc[-1]),
                xy=(s.index[-1], s.iloc[-1]),
                xytext=(8, dy),
                textcoords="offset points",
                fontsize=9,
                va="center",
                bbox=dict(boxstyle="round,pad=0.2", fc="white", ec="#BBBBBB", lw=0.8))

def rebased_100(series: pd.Series) -> pd.Series:
    s = series.dropna()
    if s.empty:
        return series * np.nan
    return 100.0 * (s / s.iloc[0])

def style_axes(ax):
    ax.grid(True, which="major", linewidth=0.8)
    for spine in ax.spines.values():
        spine.set_linewidth(0.9)

# ===========================================
# 8) CHART PACK (no stacked text anywhere)
# ===========================================

# 1) USDZAR level
fig, ax = fig_ax((14,5))
ax.plot(df.index, df["USDZAR"], linewidth=1.4)
style_axes(ax)
ax.set_ylabel("ZAR per USD")
ymin, ymax = df["USDZAR"].min(), df["USDZAR"].max()
add_regime_shading(ax, df.index, df["regime"], ymin, ymax)
pad_right(ax, df.index)
label_last(ax, df.index, df["USDZAR"], fmt="{:.2f}")
add_header(fig, ax, "USDZAR Curncy", "Weekly spot with VIX-defined regimes (shaded)", legend_ncol=1)
plt.show()

# 2) VIX
fig, ax = fig_ax((14,4))
ax.plot(df.index, df["VIX"], linewidth=1.2)
style_axes(ax)
ax.set_ylabel("Level")
ymin, ymax = df["VIX"].min(), df["VIX"].max()
add_regime_shading(ax, df.index, df["regime"], ymin, ymax)
pad_right(ax, df.index)
label_last(ax, df.index, df["VIX"], fmt="{:.2f}")
add_header(fig, ax, "VIX Index", "Weekly level (regimes defined from VIX quantiles)", legend_ncol=1)
plt.show()

# 3) DXY
fig, ax = fig_ax((14,4))
ax.plot(df.index, df["DXY"], linewidth=1.2)
style_axes(ax)
ax.set_ylabel("Level")
ymin, ymax = df["DXY"].min(), df["DXY"].max()
add_regime_shading(ax, df.index, df["regime"], ymin, ymax)
pad_right(ax, df.index)
label_last(ax, df.index, df["DXY"], fmt="{:.2f}")
add_header(fig, ax, "DXY Index", "Weekly level, regimes shaded", legend_ncol=1)
plt.show()

# 4) Commodities rebased 100
r_gold = rebased_100(df["GOLD"])
r_plat = rebased_100(df["PLAT"])
r_copp = rebased_100(df["COPPER"])
fig, ax = fig_ax((14,4))
ax.plot(r_gold.index, r_gold.values, linewidth=1.2, label="Gold (100)")
ax.plot(r_plat.index, r_plat.values, linewidth=1.2, label="Platinum (100)")
ax.plot(r_copp.index, r_copp.values, linewidth=1.2, label="Copper (100)")
style_axes(ax)
ax.set_ylabel("Index")
pad_right(ax, r_gold.index)
add_header(fig, ax, "Commodities (Rebased 100)", "Weekly levels rebased to compare cycle", legend_ncol=3)
plt.show()

# 5) Carry
fig, ax = fig_ax((14,4))
ax.plot(df.index, df["carry_bps"], linewidth=1.2)
style_axes(ax)
ax.set_ylabel("bps")
ymin, ymax = df["carry_bps"].min(), df["carry_bps"].max()
add_regime_shading(ax, df.index, df["regime"], ymin, ymax)
pad_right(ax, df.index)
label_last(ax, df.index, df["carry_bps"], fmt="{:.0f}")
add_header(fig, ax, "Carry State: (ZA 3M - US 3M)", "Level (bps), regimes shaded", legend_ncol=1)
plt.show()

# 6) CDS spread
fig, ax = fig_ax((14,4))
ax.plot(df.index, df["cds_spread_bps"], linewidth=1.2)
style_axes(ax)
ax.set_ylabel("bps")
ymin, ymax = df["cds_spread_bps"].min(), df["cds_spread_bps"].max()
add_regime_shading(ax, df.index, df["regime"], ymin, ymax)
pad_right(ax, df.index)
label_last(ax, df.index, df["cds_spread_bps"], fmt="{:.0f}")
add_header(fig, ax, "Sovereign Risk State: (ZA CDS - US CDS)", "Level (bps), regimes shaded", legend_ncol=1)
plt.show()

# 7) Rolling correlations (52w)
win = 52
tmp = df.dropna(subset=["fx_ret","dxy_ret","gold_ret","plat_ret","copp_ret"]).copy()
tmp["metals_ret"] = tmp["gold_ret"] + tmp["plat_ret"]
roll_corr_dxy = tmp["fx_ret"].rolling(win).corr(tmp["dxy_ret"])
roll_corr_met = tmp["fx_ret"].rolling(win).corr(tmp["metals_ret"])
roll_corr_cu  = tmp["fx_ret"].rolling(win).corr(tmp["copp_ret"])

fig, ax = fig_ax((14,4))
ax.plot(roll_corr_dxy.index, roll_corr_dxy.values, linewidth=1.2, label="Corr(USDZAR, DXY)")
ax.plot(roll_corr_met.index, roll_corr_met.values, linewidth=1.2, label="Corr(USDZAR, Metals)")
ax.plot(roll_corr_cu.index,  roll_corr_cu.values,  linewidth=1.2, label="Corr(USDZAR, Copper)")
style_axes(ax)
ax.set_ylabel("Correlation")
pad_right(ax, roll_corr_dxy.index)
add_header(fig, ax, "Rolling Correlations (52w)", "Weekly log returns", legend_ncol=3)
plt.show()

# 8) Rolling betas (key)
beta_plot = pd.DataFrame(index=beta_ts.index)
beta_plot["beta_DXY_total"] = beta_ts["dxy_ret"] + beta_ts["dxy_x_ro"]
beta_plot["beta_Copper"]    = beta_ts["copp_ret"]
beta_plot["beta_Metals"]    = beta_ts["gold_ret"] + beta_ts["plat_ret"]
beta_plot["beta_SA_Risk"]   = beta_ts["cds_lag"] + beta_ts["cds_x_ro"]
beta_plot["beta_Carry"]     = beta_ts["carry_lag"]

reg_for_roll = df["regime"].reindex(beta_plot.index)

fig, ax = fig_ax((14,4.5))
for col in beta_plot.columns:
    ax.plot(beta_plot.index, beta_plot[col].values, linewidth=1.1, label=col)
style_axes(ax)
ax.set_ylabel("Beta")
ymin, ymax = np.nanmin(beta_plot.values), np.nanmax(beta_plot.values)
add_regime_shading(ax, beta_plot.index, reg_for_roll, ymin, ymax)
pad_right(ax, beta_plot.index)
add_header(fig, ax, "Rolling Betas (52w, ex-ante)", "Estimated on prior 52 weeks, applied at week t", legend_ncol=3)
plt.show()

# 9) Rolling decomposition (13w sum)
stack_cols = ["DXY","Metals","Copper","SA_Risk","Carry","Const"]
stack = 100.0 * tape_s[stack_cols]
actual_pct = 100.0 * tape_s["Actual"]
resid_pct  = 100.0 * tape_s["Resid"]

fig, ax = fig_ax((14,5))
ax.stackplot(stack.index, [stack[c].values for c in stack_cols], labels=stack_cols, alpha=0.70)
ax.plot(actual_pct.index, actual_pct.values, linewidth=1.4, color="black", label=f"Actual ({SMOOTH_W}w sum, %)")
ax.plot(resid_pct.index,  resid_pct.values,  linewidth=1.1, color="#666666", label=f"Residual ({SMOOTH_W}w sum, %)")
style_axes(ax)
ax.set_ylabel(f"{SMOOTH_W}w % sum")
pad_right(ax, stack.index)
add_header(fig, ax, "USDZAR Decomposition (Rolling Betas)", f"{SMOOTH_W}-week rolling sum of contributions (readable tape)", legend_ncol=3)
plt.show()



```
