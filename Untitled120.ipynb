```python
# ============================================
# LGLV vs SPY: Beta, Convexity, Downside Convexity,
# Upside Convexity, Rolling Convexity, Rolling Beta,
# Drawdowns, Bloomberg Theme, Watermark
# Orca's Capital Strategies
# ============================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

# =========================================================
# Bloomberg-style theme (dark background + neon colours)
# =========================================================
plt.style.use('default')
plt.rcParams.update({
    'figure.facecolor': '#000000',
    'axes.facecolor':   '#000000',
    'axes.edgecolor':   '#444444',
    'axes.labelcolor':  '#E0E0E0',
    'xtick.color':      '#E0E0E0',
    'ytick.color':      '#E0E0E0',
    'text.color':       '#E0E0E0',
    'font.family':      'DejaVu Sans Mono',
    'font.size':        10,
    'lines.linewidth':  1.8,
    'axes.titleweight': 'bold',
    'grid.color':       '#333333',
    'grid.linestyle':   '--',
    'grid.linewidth':   0.4
})

BBG_COLORS = {
    'yellow':  '#F7E400',
    'cyan':    '#00E5FF',
    'orange':  '#FF8C00',
    'green':   '#00FF7F',
    'magenta': '#FF00FF',
    'red':     '#FF4C4C'
}

def add_watermark(fig):
    fig.text(
        0.5, 0.5,
        "Orca's Capital Strategies",
        fontsize=26,
        alpha=0.12,
        ha='center',
        va='center',
        rotation=30,
        color='#FFFFFF'
    )

# =========================================================
# Load CSVs
# =========================================================

spy_file  = "SPY Historical Data (5).csv"
lglv_file = "LGLV ETF Stock Price History.csv"

def clean_investing_csv(df):
    df = df.copy()
    df['Date'] = pd.to_datetime(df['Date'])
    df['Price'] = (
        df['Price']
        .astype(str)
        .str.replace(",", "", regex=False)
        .astype(float)
    )
    df = df.sort_values('Date').set_index('Date')
    return df[['Price']]

spy_raw  = pd.read_csv(spy_file)
lglv_raw = pd.read_csv(lglv_file)

spy  = clean_investing_csv(spy_raw)
lglv = clean_investing_csv(lglv_raw)

def logret(x):
    return np.log(x / x.shift(1))

spy['ret_spy']   = logret(spy['Price'])
lglv['ret_lglv'] = logret(lglv['Price'])

data = lglv[['ret_lglv']].join(spy[['ret_spy']], how='inner').dropna()

# =========================================================
# Static regressions
# =========================================================

X_lin = sm.add_constant(data['ret_spy'])
model_lin = sm.OLS(data['ret_lglv'], X_lin).fit()

X_quad = sm.add_constant(
    pd.DataFrame({
        'ret_spy': data['ret_spy'],
        'ret_spy_sq': data['ret_spy']**2
    })
)
model_quad = sm.OLS(data['ret_lglv'], X_quad).fit()

beta_capm  = model_lin.params['ret_spy']
beta_quad  = model_quad.params['ret_spy']
gamma_quad = model_quad.params['ret_spy_sq']

print("\n=== Static Linear Regression (LGLV on SPY) ===")
print(model_lin.summary())

print("\n=== Static Quadratic Regression (LGLV on SPY) ===")
print(model_quad.summary())
print(f"\nStatic beta:  {beta_quad:.4f}")
print(f"Static gamma: {gamma_quad:.4f}")

# =========================================================
# Downside-only convexity (SPY < 0)
# =========================================================

down = data[data['ret_spy'] < 0]
Xd = sm.add_constant(
    pd.DataFrame({
        'ret_spy': down['ret_spy'],
        'ret_spy_sq': down['ret_spy']**2
    })
)
model_down = sm.OLS(down['ret_lglv'], Xd).fit()
beta_down  = model_down.params['ret_spy']
gamma_down = model_down.params['ret_spy_sq']

print("\n=== Downside-only Convexity (SPY<0) ===")
print(model_down.summary())
print(f"\nDownside beta:  {beta_down:.4f}")
print(f"Downside gamma: {gamma_down:.4f}")

# =========================================================
# Upside-only convexity (SPY > 0)
# =========================================================

up = data[data['ret_spy'] > 0]
Xu = sm.add_constant(
    pd.DataFrame({
        'ret_spy': up['ret_spy'],
        'ret_spy_sq': up['ret_spy']**2
    })
)
model_up = sm.OLS(up['ret_lglv'], Xu).fit()
beta_up  = model_up.params['ret_spy']
gamma_up = model_up.params['ret_spy_sq']

print("\n=== Upside-only Convexity (SPY>0) ===")
print(model_up.summary())
print(f"\nUpside beta:  {beta_up:.4f}")
print(f"Upside gamma: {gamma_up:.4f}")

# =========================================================
# Rolling convexity and rolling beta
# =========================================================

window  = 52
min_obs = 10

roll_idx = []
roll_beta_full  = []
roll_gamma_full = []
roll_gamma_down = []
roll_gamma_up   = []

for i in range(window, len(data)):
    sub = data.iloc[i-window:i]
    roll_idx.append(sub.index[-1])

    Xr = sm.add_constant(
        pd.DataFrame({
            'ret_spy': sub['ret_spy'],
            'ret_spy_sq': sub['ret_spy']**2
        })
    )
    res_full = sm.OLS(sub['ret_lglv'], Xr).fit()
    roll_beta_full.append(res_full.params['ret_spy'])
    roll_gamma_full.append(res_full.params['ret_spy_sq'])

    sub_d = sub[sub['ret_spy'] < 0]
    if len(sub_d) >= min_obs:
        Xd_roll = sm.add_constant(
            pd.DataFrame({
                'ret_spy': sub_d['ret_spy'],
                'ret_spy_sq': sub_d['ret_spy']**2
            })
        )
        res_d = sm.OLS(sub_d['ret_lglv'], Xd_roll).fit()
        roll_gamma_down.append(res_d.params['ret_spy_sq'])
    else:
        roll_gamma_down.append(np.nan)

    sub_u = sub[sub['ret_spy'] > 0]
    if len(sub_u) >= min_obs:
        Xu_roll = sm.add_constant(
            pd.DataFrame({
                'ret_spy': sub_u['ret_spy'],
                'ret_spy_sq': sub_u['ret_spy']**2
            })
        )
        res_u = sm.OLS(sub_u['ret_lglv'], Xu_roll).fit()
        roll_gamma_up.append(res_u.params['ret_spy_sq'])
    else:
        roll_gamma_up.append(np.nan)

roll = pd.DataFrame({
    'beta_full':  roll_beta_full,
    'gamma_full': roll_gamma_full,
    'gamma_down': roll_gamma_down,
    'gamma_up':   roll_gamma_up
}, index=roll_idx)

# =========================================================
# Drawdowns
# =========================================================

def drawdown(prices):
    eq   = prices / prices.iloc[0]
    peak = eq.cummax()
    dd   = eq/peak - 1
    return eq, dd, dd.min()

eq_lglv, dd_lglv, mdd_lglv = drawdown(lglv['Price'])
eq_spy,  dd_spy,  mdd_spy  = drawdown(spy['Price'])

print(f"\nMax Drawdown LGLV: {mdd_lglv:.2%}")
print(f"Max Drawdown SPY:  {mdd_spy:.2%}")

# =========================================================
# PLOTS
# =========================================================

# ---- Equity Curve ----
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(eq_lglv.index, eq_lglv, color=BBG_COLORS['yellow'], label='LGLV')
ax.plot(eq_spy.index,  eq_spy,  color=BBG_COLORS['cyan'],   label='SPY')
ax.set_title("LGLV vs SPY  Normalized Equity Curve")
ax.legend()
add_watermark(fig)
plt.tight_layout()
plt.show()

# ---- Drawdowns ----
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(dd_lglv.index, dd_lglv, color=BBG_COLORS['orange'], label='LGLV')
ax.plot(dd_spy.index,  dd_spy,  color=BBG_COLORS['red'],    label='SPY')
ax.set_title("LGLV vs SPY  Drawdowns")
ax.legend()
add_watermark(fig)
plt.tight_layout()
plt.show()

# ---- Rolling Beta ----
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(roll.index, roll['beta_full'], color=BBG_COLORS['green'], label='Rolling Beta')
ax.axhline(beta_capm, color=BBG_COLORS['magenta'], linestyle='--', label='Static Beta')
ax.set_title("Rolling Beta (Quadratic Regression)")
ax.legend()
add_watermark(fig)
plt.tight_layout()
plt.show()

# ---- Rolling Convexity ----
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(roll.index, roll['gamma_full'], color=BBG_COLORS['yellow'],  label='Full Gamma')
ax.plot(roll.index, roll['gamma_down'], color=BBG_COLORS['red'],     label='Downside Gamma')
ax.plot(roll.index, roll['gamma_up'],   color=BBG_COLORS['cyan'],    label='Upside Gamma')
ax.axhline(0, color='#666666', linewidth=1, linestyle=':')
ax.set_title("Rolling Convexity  Full, Downside, Upside")
ax.legend()
add_watermark(fig)
plt.tight_layout()
plt.show()

# ---- Scatter (Full) ----
fig, ax = plt.subplots(figsize=(6,6))
ax.scatter(data['ret_spy'], data['ret_lglv'], s=10, color=BBG_COLORS['yellow'], alpha=0.6)

x_vals = np.linspace(data['ret_spy'].min(), data['ret_spy'].max(), 200)
y_vals = (
    model_quad.params['const']
    + model_quad.params['ret_spy']   * x_vals
    + model_quad.params['ret_spy_sq']*(x_vals**2)
)
ax.plot(x_vals, y_vals, color=BBG_COLORS['cyan'])

ax.set_title("LGLV vs SPY  Convexity (Full Sample)")
add_watermark(fig)
plt.tight_layout()
plt.show()

# ---- Scatter (Downside) ----
fig, ax = plt.subplots(figsize=(6,6))
ax.scatter(down['ret_spy'], down['ret_lglv'], s=10, color=BBG_COLORS['red'], alpha=0.6)

x_d = np.linspace(down['ret_spy'].min(), down['ret_spy'].max(), 200)
y_d = (
    model_down.params['const']
    + model_down.params['ret_spy']   * x_d
    + model_down.params['ret_spy_sq']*(x_d**2)
)
ax.plot(x_d, y_d, color=BBG_COLORS['yellow'])

ax.set_title("LGLV vs SPY  Downside Convexity")
add_watermark(fig)
plt.tight_layout()
plt.show()




```
