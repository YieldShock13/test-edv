```python
# Rolling correlation (LQD vs IEF) + realized vol (COMBINED on one chart), weekly data
# - Full fresh script (single run)
# - Local filenames only
# - Bloomberg-style dark theme + watermark "Orca's Capital Strategies"
# - Final data point labeled on ALL lines

import os
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import font_manager as fm

# =============================
# CONFIG
# =============================
LQD_CSV = "LQD Historical Data.csv"
IEF_CSV = "IEF Historical Data (4).csv"

ROLL_WEEKS = 52
ANN_FACTOR = np.sqrt(52)

WATERMARK_TEXT = "Orca's Capital Strategies"

FONT_PATH = None          # optional local font file
USE_BBG_STYLE = True

OUTPUT_DIR = "output"
OUTPUT_CSV = os.path.join(OUTPUT_DIR, "LQD_IEF_rolling_corr_rvol_52w.csv")

# =============================
# STYLE
# =============================
def setup_style():
    if FONT_PATH and os.path.exists(FONT_PATH):
        fm.fontManager.addfont(FONT_PATH)
        font_name = fm.FontProperties(fname=FONT_PATH).get_name()
    else:
        font_name = "DejaVu Sans Mono"

    mpl.rcParams["font.family"] = font_name
    mpl.rcParams["axes.titleweight"] = "bold"

    if USE_BBG_STYLE:
        mpl.rcParams["figure.facecolor"] = "black"
        mpl.rcParams["axes.facecolor"] = "black"
        mpl.rcParams["axes.edgecolor"] = "white"
        mpl.rcParams["axes.labelcolor"] = "white"
        mpl.rcParams["xtick.color"] = "white"
        mpl.rcParams["ytick.color"] = "white"
        mpl.rcParams["text.color"] = "white"
        mpl.rcParams["grid.color"] = "white"
        mpl.rcParams["grid.alpha"] = 0.15
        mpl.rcParams["axes.grid"] = True

def add_watermark(fig, text=WATERMARK_TEXT, alpha=0.08, fontsize=28):
    fig.text(
        0.5, 0.5, text,
        ha="center", va="center",
        fontsize=fontsize,
        alpha=alpha,
        rotation=22,
        transform=fig.transFigure
    )

def format_axes(ax, title: str, y_label: str = ""):
    ax.set_title(title, pad=10)
    ax.set_xlabel("")
    ax.set_ylabel(y_label)
    ax.grid(True)
    for spine in ax.spines.values():
        spine.set_alpha(0.6)

def label_last_point(ax, x, y, label, dy=0):
    x_last = x.iloc[-1]
    y_last = y.iloc[-1]
    ax.plot([x_last], [y_last], marker="o", markersize=5)
    ax.annotate(
        f"{label}: {y_last:.2%}",
        xy=(x_last, y_last),
        xytext=(8, dy),
        textcoords="offset points",
        ha="left",
        va="center",
        fontsize=10,
        alpha=0.95
    )

# =============================
# DATA LOAD
# =============================
def parse_weekly_csv(path: str, ticker: str) -> pd.DataFrame:
    df = pd.read_csv(path)
    df["Date"] = pd.to_datetime(df["Date"], format="%Y/%m/%d", errors="coerce")
    df = df.dropna(subset=["Date"])

    df["Price"] = (
        df["Price"].astype(str)
        .str.replace(",", "", regex=False)
        .str.strip()
    )
    df["Price"] = pd.to_numeric(df["Price"], errors="coerce")
    df = df.dropna(subset=["Price"])

    df = df[["Date", "Price"]].rename(columns={"Price": f"{ticker}_close"})
    df = df.sort_values("Date").drop_duplicates("Date").reset_index(drop=True)
    return df

setup_style()

lqd = parse_weekly_csv(LQD_CSV, "LQD")
ief = parse_weekly_csv(IEF_CSV, "IEF")

df = pd.merge(lqd, ief, on="Date", how="inner").sort_values("Date").reset_index(drop=True)

# =============================
# METRICS
# =============================
df["LQD_ret"] = df["LQD_close"].pct_change()
df["IEF_ret"] = df["IEF_close"].pct_change()

df["roll_corr_52w"] = df["LQD_ret"].rolling(ROLL_WEEKS).corr(df["IEF_ret"])
df["LQD_rvol_52w"] = df["LQD_ret"].rolling(ROLL_WEEKS).std() * ANN_FACTOR
df["IEF_rvol_52w"] = df["IEF_ret"].rolling(ROLL_WEEKS).std() * ANN_FACTOR

plot_df = df.dropna(subset=["roll_corr_52w", "LQD_rvol_52w", "IEF_rvol_52w"])

# =============================
# PLOT 1: ROLLING CORRELATION (SEPARATE)
# =============================
fig1, ax1 = plt.subplots(figsize=(12, 5))
ax1.plot(plot_df["Date"], plot_df["roll_corr_52w"], linewidth=1.6)
ax1.axhline(0, linewidth=1.0, alpha=0.4)
format_axes(ax1, "LQD vs IEF  52W Rolling Correlation (Weekly Returns)", "Correlation")
ax1.set_ylim(-1, 1)
label_last_point(ax1, plot_df["Date"], plot_df["roll_corr_52w"], "Last Corr", dy=0)
add_watermark(fig1)
plt.tight_layout()

# =============================
# PLOT 2: REALIZED VOL (LQD + IEF TOGETHER)
# =============================
fig2, ax2 = plt.subplots(figsize=(12, 5))

ax2.plot(plot_df["Date"], plot_df["LQD_rvol_52w"], linewidth=1.6, label="LQD")
ax2.plot(plot_df["Date"], plot_df["IEF_rvol_52w"], linewidth=1.6, label="IEF")

format_axes(ax2, "LQD vs IEF  52W Realized Vol (Annualized, Weekly Returns)", "Vol (ann.)")

label_last_point(ax2, plot_df["Date"], plot_df["LQD_rvol_52w"], "LQD", dy=6)
label_last_point(ax2, plot_df["Date"], plot_df["IEF_rvol_52w"], "IEF", dy=-6)

ax2.legend(frameon=False, loc="upper left")
add_watermark(fig2)
plt.tight_layout()

plt.show()

# =============================
# EXPORT
# =============================
os.makedirs(OUTPUT_DIR, exist_ok=True)

df[
    [
        "Date",
        "LQD_close",
        "IEF_close",
        "LQD_ret",
        "IEF_ret",
        "roll_corr_52w",
        "LQD_rvol_52w",
        "IEF_rvol_52w",
    ]
].to_csv(OUTPUT_CSV, index=False)

print("Saved:", OUTPUT_CSV)

```
