```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller

# =========================================================
# CONFIG
# =========================================================
XAU_FILE = "XAU_USD Historical Data (10).csv"
XAG_FILE = "XAG_USD Historical Data (3).csv"

MANUAL_DATE = "2026-02-10"
MANUAL_GOLD = 5035.0
MANUAL_SILVER = 80.0

WEEKLY_ANCHOR = "W-FRI"

# Z-score lookbacks to test for "suitable levels"
Z_LOOKBACKS = [52, 104, 156]   # 1y, 2y, 3y rolling mean/std on weekly ratio

# Entry/exit grids to scan (negative tail; you can add more)
ENTRY_LEVELS = [-1.5, -2.0, -2.5, -3.0]
EXIT_LEVELS  = [-0.5, -1.0, -1.5, -2.0]  # exit must be > entry (less negative)

# Your specific strategy settings
BASE_ENTRY = -2.0
BASE_EXIT  = -1.5
BLOWOUT    = -2.5
REDEPLOY_RESET_EXIT = -1.5   # same exit after redeploy
NOTIONAL_BASE = 660.0
NOTIONAL_REDEPLOY = 880.0

# Regimes requested
REGIMES = {
    "2022_NOW": ("2022-01-01", None),
    "20Y_NOW":  (None, None),  # we will cut to last 20 years programmatically
}

# =========================================================
# LOAD + CLEAN
# =========================================================
def load_price_series(path: str) -> pd.Series:
    df = pd.read_csv(path)
    df["Date"] = pd.to_datetime(df["Date"])
    df["Price"] = df["Price"].astype(str).str.replace(",", "")
    df["Price"] = pd.to_numeric(df["Price"], errors="coerce")
    df = df.sort_values("Date").dropna(subset=["Price"])
    return df.set_index("Date")["Price"]

xau = load_price_series(XAU_FILE)
xag = load_price_series(XAG_FILE)

# Weekly resample
xau_w = xau.resample(WEEKLY_ANCHOR).last()
xag_w = xag.resample(WEEKLY_ANCHOR).last()

df0 = pd.concat([xau_w.rename("XAU"), xag_w.rename("XAG")], axis=1).dropna()

# Manual append (then align to weekly again)
df0.loc[pd.Timestamp(MANUAL_DATE)] = [MANUAL_GOLD, MANUAL_SILVER]
df0 = df0.resample(WEEKLY_ANCHOR).last().dropna()

df0["ratio"] = df0["XAU"] / df0["XAG"]

# =========================================================
# OU FIT (on ratio level)
# =========================================================
def ou_fit(series: pd.Series) -> dict:
    s = series.dropna()
    s_lag = s.shift(1).dropna()
    s_now = s.loc[s_lag.index]

    X = sm.add_constant(s_lag.values)
    model = sm.OLS(s_now.values, X).fit()

    a = float(model.params[0])
    b = float(model.params[1])

    # Continuous-time OU mapping (discrete AR(1))
    # x_t = a + b x_{t-1} + eps
    # kappa = -ln(b) (if 0<b<1)
    # theta = a / (1-b)
    if 0 < b < 1:
        kappa = -np.log(b)
        half_life = np.log(2) / kappa
        theta = a / (1 - b)
    else:
        kappa = np.nan
        half_life = np.nan
        theta = np.nan

    resid = model.resid
    resid_std = float(np.std(resid, ddof=1))

    # Long-run (equilibrium) std for AR(1): sigma_eq = sigma_eps / sqrt(1-b^2)
    if abs(b) < 1:
        sigma_eq = resid_std / np.sqrt(1 - b**2)
    else:
        sigma_eq = np.nan

    adf = adfuller(s.values, autolag="AIC")
    adf_p = float(adf[1])

    return {
        "a": a, "b": b,
        "kappa": float(kappa) if np.isfinite(kappa) else np.nan,
        "half_life_weeks": float(half_life) if np.isfinite(half_life) else np.nan,
        "theta": float(theta) if np.isfinite(theta) else np.nan,
        "sigma_eps": resid_std,
        "sigma_eq": float(sigma_eq) if np.isfinite(sigma_eq) else np.nan,
        "adf_p": adf_p
    }

# =========================================================
# Z-SCORE (rolling)
# =========================================================
def add_rolling_z(dfin: pd.DataFrame, lookback: int) -> pd.DataFrame:
    df = dfin.copy()
    m = df["ratio"].rolling(lookback).mean()
    s = df["ratio"].rolling(lookback).std()
    df["z"] = (df["ratio"] - m) / s
    return df.dropna()

# =========================================================
# TRANSITION TESTS
# =========================================================
def transition_stats(df: pd.DataFrame, entry: float, exit_level: float, side: str = "neg") -> dict:
    """
    For negative side:
      entry condition: z <= entry  (entry is negative)
      exit condition:  z >= exit_level (exit_level is less negative, e.g. -1.5)
    """
    z = df["z"].values
    idx = df.index

    hit_count = 0
    total_count = 0
    tte = []  # time-to-exit in weeks

    i = 0
    n = len(z)

    while i < n:
        if z[i] <= entry:
            total_count += 1
            # from this trigger, measure time until z >= exit_level
            j = i + 1
            while j < n and z[j] < exit_level:
                j += 1
            if j < n:
                hit_count += 1
                tte.append(j - i)
                i = j  # move forward to avoid overlapping triggers
            else:
                break
        else:
            i += 1

    prob = hit_count / total_count if total_count > 0 else np.nan
    return {
        "entry": entry,
        "exit": exit_level,
        "triggers": total_count,
        "hits": hit_count,
        "p_hit": prob,
        "tte_median_weeks": float(np.median(tte)) if len(tte) else np.nan,
        "tte_mean_weeks": float(np.mean(tte)) if len(tte) else np.nan,
    }

# =========================================================
# TRADE ENGINE: YOUR STOP-AND-REDEPLOY LOGIC
# Long Gold / Short Silver when z negative
# PnL proxy uses ratio move: pnl = (ratio - entry_ratio)/entry_ratio * notional
# =========================================================
def simulate_stop_redeploy(df: pd.DataFrame,
                           base_entry: float, base_exit: float,
                           blowout: float,
                           n_base: float, n_redeploy: float) -> dict:
    z = df["z"].values
    ratio = df["ratio"].values
    dates = df.index

    in_pos = False
    notional = 0.0
    entry_ratio = np.nan
    entry_date = None
    stage = None  # "BASE" or "REDEPLOY"

    equity = 0.0
    equity_curve = []
    trades = []

    # Track MAE/MFE per trade
    cur_mae = 0.0
    cur_mfe = 0.0
    cur_entry_equity = 0.0

    for i in range(len(df)):
        zi = z[i]
        ri = ratio[i]
        dt = dates[i]

        # mark-to-market if in position (for MAE/MFE only)
        if in_pos:
            mtm = (ri - entry_ratio) / entry_ratio * notional
            # excursion relative to equity at entry of this trade-stage
            exc = mtm
            cur_mae = min(cur_mae, exc)
            cur_mfe = max(cur_mfe, exc)

        # ENTRY BASE
        if (not in_pos) and (zi <= base_entry):
            in_pos = True
            notional = n_base
            entry_ratio = ri
            entry_date = dt
            stage = "BASE"
            cur_mae = 0.0
            cur_mfe = 0.0
            cur_entry_equity = equity

        # BLOWOUT -> CLOSE BASE THEN REOPEN REDEPLOY
        elif in_pos and stage == "BASE" and (zi <= blowout):
            # close BASE
            pnl = (ri - entry_ratio) / entry_ratio * notional
            equity += pnl
            trades.append({
                "stage": "BASE_CLOSED_ON_BLOWOUT",
                "entry_date": entry_date,
                "exit_date": dt,
                "entry_z": base_entry,
                "exit_z": zi,
                "entry_ratio": entry_ratio,
                "exit_ratio": ri,
                "notional": notional,
                "pnl": pnl,
                "mae": cur_mae,
                "mfe": cur_mfe,
                "hold_weeks": int((dt - entry_date).days / 7)
            })

            # reopen REDEPLOY immediately
            in_pos = True
            notional = n_redeploy
            entry_ratio = ri
            entry_date = dt
            stage = "REDEPLOY"
            cur_mae = 0.0
            cur_mfe = 0.0
            cur_entry_equity = equity

        # EXIT
        elif in_pos and (zi >= base_exit):
            pnl = (ri - entry_ratio) / entry_ratio * notional
            equity += pnl
            trades.append({
                "stage": stage,
                "entry_date": entry_date,
                "exit_date": dt,
                "entry_z": None,
                "exit_z": zi,
                "entry_ratio": entry_ratio,
                "exit_ratio": ri,
                "notional": notional,
                "pnl": pnl,
                "mae": cur_mae,
                "mfe": cur_mfe,
                "hold_weeks": int((dt - entry_date).days / 7)
            })
            in_pos = False
            notional = 0.0
            entry_ratio = np.nan
            entry_date = None
            stage = None
            cur_mae = 0.0
            cur_mfe = 0.0

        equity_curve.append(equity)

    eq = pd.Series(equity_curve, index=df.index)
    dd = (eq - eq.cummax()).min()

    trades_df = pd.DataFrame(trades)
    return {
        "trades": trades_df,
        "equity": eq,
        "max_drawdown": float(dd)
    }

# =========================================================
# THRESHOLD SCAN ("find suitable Z levels")
# =========================================================
def scan_thresholds(df: pd.DataFrame):
    results = []
    for entry in ENTRY_LEVELS:
        for exit_level in EXIT_LEVELS:
            if exit_level <= entry:
                continue
            # simple: enter when z<=entry, exit when z>=exit_level (no blowout logic here)
            # used only to assess signal frequency + basic expectancy
            z = df["z"].values
            ratio = df["ratio"].values
            in_pos = False
            entry_ratio = np.nan
            pnls = []
            holds = []

            for i in range(len(df)):
                if (not in_pos) and (z[i] <= entry):
                    in_pos = True
                    entry_ratio = ratio[i]
                    entry_i = i
                elif in_pos and (z[i] >= exit_level):
                    pnl = (ratio[i] - entry_ratio) / entry_ratio
                    pnls.append(pnl)
                    holds.append(i - entry_i)
                    in_pos = False

            if len(pnls) >= 1:
                results.append({
                    "entry": entry,
                    "exit": exit_level,
                    "trades": len(pnls),
                    "win_rate": float(np.mean(np.array(pnls) > 0)),
                    "mean_pnl_ratio": float(np.mean(pnls)),
                    "median_hold_weeks": float(np.median(holds)),
                })
            else:
                results.append({
                    "entry": entry,
                    "exit": exit_level,
                    "trades": 0,
                    "win_rate": np.nan,
                    "mean_pnl_ratio": np.nan,
                    "median_hold_weeks": np.nan,
                })

    out = pd.DataFrame(results).sort_values(["trades", "entry"], ascending=[False, True])
    return out

# =========================================================
# RUN EVERYTHING FOR EACH REGIME + EACH Z LOOKBACK
# =========================================================
def cut_to_last_20y(df: pd.DataFrame) -> pd.DataFrame:
    end = df.index.max()
    start = end - pd.DateOffset(years=20)
    return df[df.index >= start]

def plot_basic_z(df: pd.DataFrame, title: str):
    plt.figure(figsize=(14,5))
    plt.plot(df.index, df["z"])
    for lvl in [0, -1.5, -2.0, -2.5, 2.0]:
        plt.axhline(lvl, linestyle="--" if lvl != 0 else "-", linewidth=1)
    plt.title(title)
    plt.grid(alpha=0.3)
    plt.show()

def plot_density(series: pd.Series, title: str, bins: int = 30):
    plt.figure(figsize=(7,4))
    plt.hist(series.dropna().values, bins=bins, density=True)
    plt.title(title)
    plt.grid(alpha=0.3)
    plt.show()

for regime_name, (start_date, end_date) in REGIMES.items():
    # slice raw ratio frame
    dreg = df0.copy()
    if regime_name == "20Y_NOW":
        dreg = cut_to_last_20y(dreg)
    if start_date is not None:
        dreg = dreg[dreg.index >= pd.Timestamp(start_date)]
    if end_date is not None:
        dreg = dreg[dreg.index <= pd.Timestamp(end_date)]

    print("\n" + "="*70)
    print(f"REGIME: {regime_name}")
    print(f"Date range: {dreg.index.min().date()} -> {dreg.index.max().date()} | obs={len(dreg)}")
    print("="*70)

    # OU + ADF on ratio level (regime raw)
    ou = ou_fit(dreg["ratio"])
    print("OU (ratio)  a:", round(ou["a"],6), " b:", round(ou["b"],6))
    print("kappa:", round(ou["kappa"],6), " half-life (weeks):", round(ou["half_life_weeks"],2))
    print("theta:", round(ou["theta"],6), " sigma_eq:", round(ou["sigma_eq"],6))
    print("ADF p-value:", round(ou["adf_p"],4))

    for lookback in Z_LOOKBACKS:
        dz = add_rolling_z(dreg, lookback)
        if len(dz) < 200:
            print(f"\n[Skip lookback={lookback}] too few obs after rolling dropna: {len(dz)}")
            continue

        print("\n" + "-"*70)
        print(f"Z LOOKBACK: {lookback} weeks")
        print("Current Z:", round(dz["z"].iloc[-1],3))
        print("-"*70)

        # 1) Z time series plot
        plot_basic_z(dz, f"{regime_name} | Z-score (rolling {lookback}w)")

        # 2) Z distribution density
        plot_density(dz["z"], f"{regime_name} | Z density | lookback={lookback}")

        # 3) Threshold scan to "find suitable Z scores"
        scan = scan_thresholds(dz)
        print("\nTop threshold candidates by trade count (first 15 rows):")
        print(scan.head(15).to_string(index=False))

        # 4) Transition probabilities (explicitly include your requested transitions)
        t1 = transition_stats(dz, entry=-2.0, exit_level=-1.5)
        t2 = transition_stats(dz, entry=-2.5, exit_level=-2.0)
        print("\nTransition stats:")
        print(" -2.0 -> -1.5 :", t1)
        print(" -2.5 -> -2.0 :", t2)

        # 5) Your full stop-and-redeploy strategy backtest
        sim = simulate_stop_redeploy(
            dz,
            base_entry=BASE_ENTRY,
            base_exit=BASE_EXIT,
            blowout=BLOWOUT,
            n_base=NOTIONAL_BASE,
            n_redeploy=NOTIONAL_REDEPLOY
        )

        trades = sim["trades"]
        eq = sim["equity"]
        max_dd = sim["max_drawdown"]

        print("\nSTOP+REDEPLOY RESULTS (your exact rules):")
        print("Trades:", len(trades))
        if len(trades) > 0:
            win_rate = float(np.mean(trades["pnl"].values > 0))
            print("Win rate:", round(win_rate,3))
            print("Total PnL (USD proxy):", round(float(trades["pnl"].sum()),2))
            print("Mean PnL:", round(float(trades["pnl"].mean()),2))
            print("Median PnL:", round(float(trades["pnl"].median()),2))
            print("Median hold (weeks):", float(trades["hold_weeks"].median()))
            print("Max Drawdown (USD):", round(max_dd,2))
        else:
            print("No trades under these rules in this regime/lookback.")

        # 6) Equity curve + drawdown visual
        plt.figure(figsize=(14,4))
        plt.plot(eq.index, eq.values)
        plt.title(f"Equity Curve | {regime_name} | lookback={lookback}")
        plt.grid(alpha=0.3)
        plt.show()

        # 7) PnL density, holding density, MAE/MFE density (if enough trades)
        if len(trades) > 0:
            plot_density(trades["pnl"], f"PnL density | {regime_name} | lookback={lookback}", bins=20)
            plot_density(trades["hold_weeks"], f"Holding (weeks) density | {regime_name} | lookback={lookback}", bins=20)
            plot_density(trades["mae"], f"MAE density (USD) | {regime_name} | lookback={lookback}", bins=20)
            plot_density(trades["mfe"], f"MFE density (USD) | {regime_name} | lookback={lookback}", bins=20)

            print("\nTrade list (all trades):")
            cols = ["stage","entry_date","exit_date","notional","pnl","mae","mfe","hold_weeks","exit_z"]
            print(trades[cols].to_string(index=False))





```
