```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.stats.anova import anova_lm
from statsmodels.stats.diagnostic import het_breuschpagan, acorr_ljungbox
from statsmodels.stats.outliers_influence import variance_inflation_factor
from fredapi import Fred

# === API Setup ===
fred = Fred(api_key='680738fcd07f4b0c379a0a46a4e6afdc')

# === FRED series dictionary ===
series = {
    'UST10Y': 'GS10',
    'UST2Y': 'GS2',
    'FedFunds': 'FEDFUNDS',
    'TP_KW10Y': 'THREEFYTP10',
    'InflExp_5Y5Y': 'T5YIFR',
    'UnempRate': 'UNRATE',
    'CPI': 'CPIAUCNS',
    'GDP': 'GDPC1',
    'IndProd': 'INDPRO',
    'VIX': 'VIXCLS',
    'USD_Index': 'DTWEXBGS'
}

# === Pull and format data ===
def get_series(sid):
    data = fred.get_series(sid)
    return data.resample('QE').last()

df = pd.DataFrame({k: get_series(v) for k, v in series.items()})
df = df.loc['1980':]

# === Derived YoY variables ===
df['CPI_YoY'] = df['CPI'].pct_change(4, fill_method=None) * 100
df['GDP_YoY'] = df['GDP'].pct_change(4, fill_method=None) * 100
df['IndProd_YoY'] = df['IndProd'].pct_change(4, fill_method=None) * 100
df['USD_YoY'] = df['USD_Index'].pct_change(4, fill_method=None) * 100
df.drop(columns=['CPI', 'GDP', 'IndProd', 'USD_Index'], inplace=True)
df.dropna(inplace=True)

# === Regression using formula API ===
model = smf.ols("UST10Y ~ FedFunds + TP_KW10Y + InflExp_5Y5Y + UnempRate + GDP_YoY + CPI_YoY + VIX + USD_YoY", data=df).fit()

# === Add predictions/residuals ===
df['Predicted'] = model.predict(df)
df['Residuals'] = df['UST10Y'] - df['Predicted']

# === ANOVA Table ===
anova = anova_lm(model, typ=2)
print("\nANOVA Table:\n", anova)

# === Breusch–Pagan ===
bp = het_breuschpagan(model.resid, model.model.exog)
print("\nBreusch–Pagan test:\n", dict(zip(['LM', 'p-value', 'F', 'F p-value'], bp)))

# === Ljung–Box ===
print("\nLjung–Box test:\n", acorr_ljungbox(model.resid, lags=[4], return_df=True))

# === VIFs (exclude intercept) ===
X = model.model.exog
vif = pd.DataFrame()
vif["Feature"] = model.model.exog_names
vif["VIF"] = [variance_inflation_factor(X, i) for i in range(X.shape[1])]
print("\nVIF:\n", vif)

# === Plot: Actual vs Predicted (Only Label Q2 2025) ===
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['UST10Y'], label='Actual 10Y Yield', marker='o')
plt.plot(df.index, df['Predicted'], label='Predicted 10Y Yield', linestyle='--', marker='x')

# Label only the last date
last_date = df.index[-1]
plt.text(last_date, df['UST10Y'].iloc[-1] + 0.1, f"{df['UST10Y'].iloc[-1]:.2f}", color='blue')
plt.text(last_date, df['Predicted'].iloc[-1] - 0.2, f"{df['Predicted'].iloc[-1]:.2f}", color='black')

plt.title('Actual vs Predicted U.S. 10Y Yield (Quarterly)')
plt.xlabel('Date')
plt.ylabel('Yield (%)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# === Plot: Residuals over time ===
plt.figure(figsize=(10, 4))
plt.plot(df.index, df['Residuals'], label='Residuals')
plt.axhline(0, color='red', linestyle='--')
plt.title('Model Residuals Over Time')
plt.grid(True)
plt.tight_layout()
plt.show()

# === Residual distribution ===
plt.figure(figsize=(6, 4))
sns.histplot(df['Residuals'], kde=True, bins=20)
plt.title('Distribution of Residuals')
plt.grid(True)
plt.tight_layout()
plt.show()

# === Residuals vs Fitted ===
plt.figure(figsize=(6, 4))
plt.scatter(df['Predicted'], df['Residuals'], alpha=0.7)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Fitted Yield')
plt.ylabel('Residuals')
plt.title('Residuals vs Fitted Values')
plt.grid(True)
plt.tight_layout()
plt.show()








```
