```python
# ============================================================
# Gold & Silver (FULL HISTORY):
# 1) Z-score of Silver/Gold ratio (final point labeled)
# 2) Drawdowns for BOTH, with max DD annotations (value+date)
# 3) Rolling correlation of daily returns (final value labeled)
# 4) Annual return football fields (expanded height; negatives red)
# Watermark: Orca's Capital Strategies
# ============================================================

import pandas as pd, numpy as np, matplotlib.pyplot as plt, glob

# -----------------------------
# Style / Config
# -----------------------------
plt.style.use("dark_background")
BG        = "#0b1e2d"
GOLD_COL  = "#f2c14e"
SILV_COL  = "#c0c0c0"
CYAN_COL  = "#17becf"
WATERMARK = "Orca's Capital Strategies"

Z_WIN     = 252             # rolling window for z-score (trading days ~ 1y)
CORR_WIN  = 90              # rolling window for correlation (~3m)

# -----------------------------
# Loaders
# -----------------------------
def load_investing_price(pattern, label):
    paths = sorted(glob.glob(pattern))
    if not paths:
        raise FileNotFoundError(f"No file matched: {pattern}")
    df = pd.read_csv(paths[0])
    if "Date" not in df.columns or "Price" not in df.columns:
        raise ValueError(f"{paths[0]} must contain 'Date' and 'Price' columns.")
    df["Date"]  = pd.to_datetime(df["Date"])
    df["Price"] = pd.to_numeric(df["Price"].astype(str).str.replace(",", "", regex=False), errors="coerce")
    s = pd.Series(df["Price"].values, index=df["Date"], name=label).sort_index().dropna()
    return s

gold   = load_investing_price("XAU_USD Historical Data*.csv", "Gold")
silver = load_investing_price("XAG_USD Historical Data*.csv", "Silver")

# Use UNION of dates (keep full histories)
all_idx = gold.index.union(silver.index).sort_values()
gold    = gold.reindex(all_idx)
silver  = silver.reindex(all_idx)

# -----------------------------
# Helper functions
# -----------------------------
def ocean_axes(ax, title):
    ax.set_facecolor(BG); ax.figure.set_facecolor(BG)
    ax.set_title(title, fontsize=15, fontweight="bold", color="white", pad=14)
    ax.tick_params(colors="white")
    ax.grid(alpha=0.25, color="white", ls="--")
    for spine in ["bottom","left"]:
        ax.spines[spine].set_color("white")
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

def add_watermark(ax):
    ax.text(0.5, 0.05, WATERMARK, transform=ax.transAxes,
            fontsize=14, color="white", alpha=0.15,
            ha="center", va="center", rotation=25)

def drawdown_series(price):
    """Peak-to-trough drawdown series (returns as negative values)."""
    px = price.dropna()
    if px.empty: return pd.Series(dtype=float)
    eq = px / px.iloc[0]
    peak = eq.cummax()
    dd = eq / peak - 1.0
    return dd

def annual_returns_from_daily(px):
    """Calendar-year total return using first/last price of each year."""
    s = px.dropna()
    if s.empty: return pd.Series(dtype=float)
    g = s.groupby(s.index.year)
    yr = g.apply(lambda v: (v.iloc[-1] / v.iloc[0] - 1.0) * 100.0)
    yr.index.name = "Year"
    return yr

# -----------------------------
# 1) Z-score of Silver/Gold ratio
# -----------------------------
ratio = silver / gold
z = (ratio - ratio.rolling(Z_WIN, min_periods=int(Z_WIN*0.2)).mean()) / \
    ratio.rolling(Z_WIN, min_periods=int(Z_WIN*0.2)).std(ddof=0)

fig, ax = plt.subplots(figsize=(12,6))
ax.plot(z.index, z.values, color=CYAN_COL, lw=1.8, label="Z-Score (Silver  Gold)")
ax.axhline(0,  color="white", lw=1, alpha=0.6)
ax.axhline(1,  color="white", lw=0.8, ls="--", alpha=0.5)
ax.axhline(-1, color="white", lw=0.8, ls="--", alpha=0.5)
ax.axhline(2,  color="white", lw=0.8, ls=":",  alpha=0.5)
ax.axhline(-2, color="white", lw=0.8, ls=":",  alpha=0.5)
# Final point label
if not z.dropna().empty:
    last_date, last_val = z.dropna().index[-1], z.dropna().iloc[-1]
    ax.scatter(last_date, last_val, color="red", s=40, zorder=5)
    ax.annotate(f"{last_val:.2f}", xy=(last_date, last_val),
                xytext=(8,6), textcoords="offset points",
                color="white", fontsize=9, weight="bold")
ocean_axes(ax, "Z-Score of Silver/Gold Price Ratio (Full History)")
ax.set_ylabel("Z-Score ()", color="white")
ax.legend(facecolor=BG, edgecolor="none", fontsize=9)
add_watermark(ax)
plt.tight_layout(); plt.show()

# -----------------------------
# 2) Drawdowns for BOTH (with max DD annotations)
# -----------------------------
dd_gold   = drawdown_series(gold)
dd_silver = drawdown_series(silver)

mdd_g_val = dd_gold.min() if not dd_gold.empty else np.nan
mdd_s_val = dd_silver.min() if not dd_silver.empty else np.nan
mdd_g_pct = mdd_g_val * 100 if pd.notna(mdd_g_val) else np.nan
mdd_s_pct = mdd_s_val * 100 if pd.notna(mdd_s_val) else np.nan
mdd_g_dt  = dd_gold.idxmin() if not dd_gold.empty else None
mdd_s_dt  = dd_silver.idxmin() if not dd_silver.empty else None

fig, ax = plt.subplots(figsize=(12,6))
ax.plot(dd_gold.index, dd_gold.values*100,   color=GOLD_COL, lw=1.8, label="Gold drawdown")
ax.plot(dd_silver.index, dd_silver.values*100, color=SILV_COL,  lw=1.8, label="Silver drawdown")

# Annotations for BOTH minima
if pd.notna(mdd_g_pct):
    ax.scatter(mdd_g_dt, mdd_g_pct, color=GOLD_COL, s=45, zorder=5)
    ax.annotate(f"Gold MDD: {mdd_g_pct:.1f}% on {mdd_g_dt.date()}",
                xy=(mdd_g_dt, mdd_g_pct), xytext=(10,10),
                textcoords="offset points", color=GOLD_COL, fontsize=9, weight="bold")
if pd.notna(mdd_s_pct):
    ax.scatter(mdd_s_dt, mdd_s_pct, color=SILV_COL, s=45, zorder=5)
    ax.annotate(f"Silver MDD: {mdd_s_pct:.1f}% on {mdd_s_dt.date()}",
                xy=(mdd_s_dt, mdd_s_pct), xytext=(10,-16),
                textcoords="offset points", color=SILV_COL, fontsize=9, weight="bold")

ocean_axes(ax, "Drawdowns from Peak  Gold vs Silver (Full History)")
ax.set_ylabel("Drawdown (%)", color="white")
ax.legend(facecolor=BG, edgecolor="none", fontsize=9)
add_watermark(ax)
plt.tight_layout(); plt.show()

# -----------------------------
# 3) Rolling correlation of daily returns (final value labeled)
# -----------------------------
ret_gold   = gold.pct_change()
ret_silver = silver.pct_change()
ret_df = pd.concat([ret_gold, ret_silver], axis=1)
ret_df.columns = ["Gold_ret", "Silver_ret"]

roll_corr = ret_df["Gold_ret"].rolling(CORR_WIN, min_periods=int(CORR_WIN*0.5)) \
                            .corr(ret_df["Silver_ret"])

fig, ax = plt.subplots(figsize=(12,6))
ax.plot(roll_corr.index, roll_corr.values, color=CYAN_COL, lw=1.8, label=f"Rolling Corr ({CORR_WIN}d)")
ax.axhline(0, color="white", lw=1, alpha=0.6)
if not roll_corr.dropna().empty:
    c_last_date, c_last_val = roll_corr.dropna().index[-1], roll_corr.dropna().iloc[-1]
    ax.scatter(c_last_date, c_last_val, color="red", s=40, zorder=5)
    ax.annotate(f"{c_last_val:.2f}", xy=(c_last_date, c_last_val),
                xytext=(8,6), textcoords="offset points",
                color="white", fontsize=9, weight="bold")
ocean_axes(ax, "Rolling Correlation of Daily Returns  Gold vs Silver")
ax.set_ylabel("Correlation", color="white")
ax.legend(facecolor=BG, edgecolor="none", fontsize=9)
add_watermark(ax)
plt.tight_layout(); plt.show()

# -----------------------------
# 4) Annual return football fields (expanded height; all years)
# -----------------------------
ann_gold   = annual_returns_from_daily(gold)     # % values
ann_silver = annual_returns_from_daily(silver)   # % values
all_years  = sorted(set(ann_gold.index) | set(ann_silver.index))

# Reindex to the union of years; if a year truly has no begin/end for an asset, keep NaN (dont fabricate 0)
ann_gold  = ann_gold.reindex(all_years)
ann_silver= ann_silver.reindex(all_years)

# dynamic height so y-axis is not cramped
rows = len(all_years)
height_per_row = 0.28   # tweak to taste
fig_height = max(8, rows * height_per_row)

fig, axes = plt.subplots(2, 1, figsize=(13, fig_height), sharex=True)

def football(ax, series, color, title):
    for y in all_years:
        v = series.get(y, np.nan)
        # Only plot if we have a real return for that year
        if pd.notna(v):
            ax.barh(y, v, color=(color if v >= 0 else "crimson"), alpha=0.95)
            # value label
            ax.text(v + (2 if v >= 0 else -2), y, f"{v:.1f}%",
                    va="center", ha=("left" if v >= 0 else "right"),
                    color="white", fontsize=8)
    ax.axvline(0, color="white", lw=1)
    ax.set_yticks(all_years)
    ax.set_ylabel("Year", color="white", fontsize=9)
    ax.set_title(title, color="white", fontsize=12, pad=8)
    ax.tick_params(colors="white")
    ax.set_facecolor(BG)

football(axes[0], ann_gold,   GOLD_COL, "Annual Returns  Gold (Calendar Years)")
football(axes[1], ann_silver, SILV_COL, "Annual Returns  Silver (Calendar Years)")
axes[1].set_xlabel("Return (%)", color="white", fontsize=10)

# Watermark across the bottom subplot
axes[1].text(0.5, -0.08, WATERMARK, transform=axes[1].transAxes,
             fontsize=14, color="white", alpha=0.15,
             ha="center", va="center", rotation=25)

plt.tight_layout()
plt.show()




```
