```python
# ==========================================================
# Monthly (1st) DE vs US  Yields, Futures & Two Strategies
# Strategy A (L/S): Futures z ONLY, thresholds  {0.0, 0.5, 1.0}
#   z >= +thr -> Long US / Short DE   (ret = US - DE)
#   z <= -thr -> Long DE / Short US   (ret = DE - US)
#   else flat
# Strategy B (unchanged): Yield-only z rule at 1 (long-only)
# Watermark: "Orca's Capital Strategies"
# ==========================================================
import os, glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import statsmodels.api as sm

# ---------------- Config ----------------
WATERMARK = "Orca's Capital Strategies"
DATE_FMT_M = DateFormatter("%Y-%m")
# For Strategy A only: which futures z to use as the signal
FUT_SIGNAL_SOURCE = "1x1"   # "1x1" uses z of (DEUS); set "hedged" to use z of (DEUS)
THRESH_LIST = [0.0, 0.5, 1.0]   # run Strategy A for these sigma thresholds

# Exact filenames
FN_DE_YLD = "Germany 2-Year Bond Yield Historical Data (2).csv"
FN_US_YLD = "United States 2-Year Bond Yield Historical Data (1).csv"
FN_DE_FUT = "Euro SCHATZ Futures Historical Data.csv"
FN_US_FUT = "US 2 Year T-Note Futures Historical Data.csv"

# ---------------- Helpers ----------------
def resolve_path(filename):
    for p in (f"/content/{filename}", f"/mnt/data/{filename}", filename):
        if os.path.isfile(p): return p
    for base in ("/content", "/mnt/data", "."):
        hits = glob.glob(os.path.join(base, "**", filename), recursive=True)
        if hits: return hits[0]
    raise FileNotFoundError(f"Could not find: {filename}")

def robust_date(s: pd.Series) -> pd.Series:
    s = s.astype(str)
    d1 = pd.to_datetime(s, dayfirst=True,  errors="coerce")
    d2 = pd.to_datetime(s, dayfirst=False, errors="coerce")
    return d1 if d1.max() >= d2.max() else d2

def parse_number(series: pd.Series) -> pd.Series:
    s = series.astype(str).str.strip().str.replace("%","",regex=False)
    smpl = s.sample(min(200, len(s)), random_state=0)
    comma_no_dot = ((smpl.str.contains(",")) & ~smpl.str.contains(r"\.")).mean()
    if comma_no_dot > 0.6:
        s = s.str.replace(".", "", regex=False)   # thousands dot
        s = s.str.replace(",", ".", regex=False)  # decimal comma  dot
    else:
        s = s.str.replace(",", "", regex=False)   # thousands comma
    return pd.to_numeric(s, errors="coerce")

def load_monthly_first(path: str, colname: str) -> pd.Series:
    df = pd.read_csv(path)
    df.columns = [c.strip() for c in df.columns]
    if "Date" not in df.columns or "Price" not in df.columns:
        raise ValueError(f"'Date' and 'Price' required in {path}. Found: {df.columns.tolist()}")
    df["Date"]  = robust_date(df["Date"])
    df["Price"] = parse_number(df["Price"])
    df = df.dropna(subset=["Date","Price"]).sort_values("Date").set_index("Date")
    df = df[~df.index.duplicated(keep="first")]  # one per month (files are stamped on 1st)
    return df["Price"].rename(colname)

def add_wm(fig): fig.text(0.01, 0.01, WATERMARK, fontsize=12, alpha=0.35, ha="left", va="bottom")
def annotate_last(ax, x, y, fmt="{:.2f}", dx=8, dy=0): ax.annotate(fmt.format(y[-1]), xy=(x[-1], y[-1]), xytext=(dx, dy), textcoords="offset points", va="center")
def finalize(fig): add_wm(fig); plt.tight_layout(); plt.show()

def compute_metrics(inc: pd.Series):
    inc = inc.astype(float).fillna(0.0)
    equity = (1 + inc).cumprod()
    roll_max = equity.cummax()
    drawdown = equity / roll_max - 1.0
    cum_gain = equity.iloc[-1] - 1.0
    wins = (inc > 0).sum(); losses = (inc < 0).sum()
    win_loss_ratio = (wins / losses) if losses > 0 else np.nan
    var95 = -np.nanpercentile(inc.replace([np.inf,-np.inf], np.nan).dropna(), 5)  # 1m VaR(95%)
    ann_vol = inc.std(ddof=0) * np.sqrt(12)
    n = len(inc)
    cagr = equity.iloc[-1]**(12 / n) - 1 if n > 0 else np.nan
    trr = (cagr / ann_vol) if (ann_vol and ann_vol != 0) else np.nan
    return {"equity":equity,"drawdown":drawdown,"cum_gain":cum_gain,"wins":wins,"losses":losses,
            "win_loss_ratio":win_loss_ratio,"var95":var95,"ann_vol":ann_vol,"cagr":cagr,"trr":trr,
            "max_dd":drawdown.min()}

# ---------------- Load data ----------------
p_de_yld = resolve_path(FN_DE_YLD)
p_us_yld = resolve_path(FN_US_YLD)
p_de_fut = resolve_path(FN_DE_FUT)
p_us_fut = resolve_path(FN_US_FUT)

DE2Y = load_monthly_first(p_de_yld, "DE2Y_yld")
US2Y = load_monthly_first(p_us_yld, "US2Y_yld")
DE_F = load_monthly_first(p_de_fut, "DE_fut")
US_F = load_monthly_first(p_us_fut, "US_fut")

# Align indices
ix_yld = DE2Y.index.intersection(US2Y.index)
yld = pd.concat([DE2Y.reindex(ix_yld), US2Y.reindex(ix_yld)], axis=1).dropna()

ix_fut = DE_F.index.intersection(US_F.index)
fut = pd.concat([DE_F.reindex(ix_fut), US_F.reindex(ix_fut)], axis=1).dropna()

# ---------------- Yield spread (bps) + z + OLS ----------------
yld["Spread_bps"] = (yld["DE2Y_yld"] - yld["US2Y_yld"]) * 100.0
mu_y, sd_y = yld["Spread_bps"].mean(), yld["Spread_bps"].std(ddof=0)
yld["Z_yield"] = (yld["Spread_bps"] - mu_y) / sd_y if sd_y != 0 else np.nan

Xy = sm.add_constant(yld["US2Y_yld"])
yy = yld["DE2Y_yld"]
ols_yield = sm.OLS(yy, Xy).fit()
beta_y = float(ols_yield.params["US2Y_yld"]); alpha_y = float(ols_yield.params["const"]); r2_y = float(ols_yield.rsquared)

# ---------------- Futures spreads + z ----------------
# 11
fut["Spread_1x1"] = fut["DE_fut"] - fut["US_fut"]
mu_1x1, sd_1x1 = fut["Spread_1x1"].mean(), fut["Spread_1x1"].std(ddof=0)
fut["Z_1x1"] = (fut["Spread_1x1"] - mu_1x1) / sd_1x1 if sd_1x1 != 0 else np.nan

# OLS-hedged (DE_fut ~ const + beta * US_fut)
Xf = sm.add_constant(fut["US_fut"])
yf = fut["DE_fut"]
ols_fut = sm.OLS(yf, Xf).fit()
beta_f = float(ols_fut.params["US_fut"]); alpha_f = float(ols_fut.params["const"])

fut["Spread_hedged"] = fut["DE_fut"] - beta_f * fut["US_fut"]
mu_h, sd_h = fut["Spread_hedged"].mean(), fut["Spread_hedged"].std(ddof=0)
fut["Z_hedged"] = (fut["Spread_hedged"] - mu_h) / sd_h if sd_h != 0 else np.nan

# ---------------- Forward returns (signal at t, hold to t+1) ----------------
DE_ret_fwd = DE_F.shift(-1) / DE_F - 1.0
US_ret_fwd = US_F.shift(-1) / US_F - 1.0
ret_df = pd.concat([DE_ret_fwd.rename("DE_ret_fwd"), US_ret_fwd.rename("US_ret_fwd")], axis=1)

# ==========================================================
# STRATEGY A  Futures-z ONLY, LONG/SHORT pair  for each THRESH
# ==========================================================
panelA_base = fut.join(ret_df, how="inner").dropna()
zsig_full = panelA_base["Z_1x1"] if FUT_SIGNAL_SOURCE.lower() == "1x1" else panelA_base["Z_hedged"]

def run_strategy_A_pair(thresh: float):
    df = panelA_base.copy()
    zsig = zsig_full.copy()

    longUS_shortDE = (zsig >= +thresh)
    longDE_shortUS = (zsig <= -thresh)

    df["Strat_ret"] = 0.0
    df.loc[longUS_shortDE, "Strat_ret"] = df.loc[longUS_shortDE, "US_ret_fwd"] - df.loc[longUS_shortDE, "DE_ret_fwd"]
    df.loc[longDE_shortUS, "Strat_ret"] = df.loc[longDE_shortUS, "DE_ret_fwd"] - df.loc[longDE_shortUS, "US_ret_fwd"]

    mets = compute_metrics(df["Strat_ret"])
    equity, drawdown = mets["equity"], mets["drawdown"]

    # Z visuals
    figZ, axZ = plt.subplots(figsize=(11,5.0))
    axZ.plot(fut.index, fut["Z_1x1"], label="Futures z (11)")
    axZ.plot(fut.index, fut["Z_hedged"], label=f"Futures z (hedged ={beta_f:.3f})")
    axZ.axhline(0, linewidth=1)
    if thresh > 0:
        axZ.axhline(+thresh, linestyle="--", alpha=0.8, label=f"+{thresh}")
        axZ.axhline(-thresh, linestyle="--", alpha=0.8, label=f"{thresh}")
    axZ.set_title(f"Strategy A (L/S)  Futures Z-Scores (threshold = {thresh})")
    axZ.set_ylabel("Z"); axZ.set_xlabel("Date"); axZ.grid(True, alpha=0.3)
    axZ.xaxis.set_major_formatter(DATE_FMT_M)
    annotate_last(axZ, fut.index, fut["Z_1x1"].values, fmt="{:.2f}")
    annotate_last(axZ, fut.index, fut["Z_hedged"].values, fmt="{:.2f}", dy=-12)
    axZ.legend(); finalize(figZ)

    # Entries on price chart
    figP, axP = plt.subplots(figsize=(12,5.6))
    axP.plot(fut.index, fut["DE_fut"], label="DE futures")
    axP.plot(fut.index, fut["US_fut"], label="US futures")
    de_bias = df.index[longDE_shortUS]; us_bias = df.index[longUS_shortDE]
    axP.scatter(de_bias, fut.loc[de_bias, "DE_fut"], marker="^", s=90, label="ENTRY: Long DE / Short US", zorder=5)
    axP.scatter(us_bias, fut.loc[us_bias, "US_fut"], marker="v", s=90, label="ENTRY: Long US / Short DE", zorder=5)
    axP.set_title(f"Strategy A (L/S) Entries  {FUT_SIGNAL_SOURCE.upper()} z, thr={thresh}")
    axP.set_ylabel("Price"); axP.set_xlabel("Date"); axP.grid(True, alpha=0.3)
    axP.xaxis.set_major_formatter(DATE_FMT_M)
    annotate_last(axP, fut.index, fut["DE_fut"].values, fmt="{:.2f}")
    annotate_last(axP, fut.index, fut["US_fut"].values, fmt="{:.2f}")
    axP.legend(); finalize(figP)

    # Equity
    figE, axE = plt.subplots(figsize=(11,5.0))
    axE.plot(df.index, equity, label="Equity")
    axE.set_title(f"Strategy A (L/S)  Equity (thr={thresh})")
    axE.set_ylabel("Growth of $1"); axE.set_xlabel("Date"); axE.grid(True, alpha=0.3)
    axE.xaxis.set_major_formatter(DATE_FMT_M)
    annotate_last(axE, df.index, equity.values, fmt="{:.3f}")
    axE.legend(); finalize(figE)

    # Monthly returns
    figR, axR = plt.subplots(figsize=(11,4.2))
    colors = np.where(df["Strat_ret"] >= 0, "tab:blue", "tab:orange")
    axR.bar(df.index, df["Strat_ret"].values, width=20, align="center", color=colors)
    axR.axhline(0, linewidth=1)
    axR.set_title(f"Strategy A (L/S)  Monthly Returns (thr={thresh})")
    axR.set_ylabel("Return"); axR.set_xlabel("Date"); axR.grid(True, alpha=0.25)
    axR.xaxis.set_major_formatter(DATE_FMT_M)
    finalize(figR)

    # Drawdown
    figD, axD = plt.subplots(figsize=(11,3.8))
    axD.plot(df.index, drawdown)
    axD.set_title(f"Strategy A (L/S)  Drawdown (thr={thresh})")
    axD.set_ylabel("Drawdown"); axD.set_xlabel("Date"); axD.grid(True, alpha=0.3)
    axD.xaxis.set_major_formatter(DATE_FMT_M)
    finalize(figD)

    # Per-trade table
    trades = df.loc[longUS_shortDE | longDE_shortUS,
                    ["DE_ret_fwd","US_ret_fwd","Strat_ret"]].copy()
    trades["Bias"] = np.where(longUS_shortDE.loc[trades.index], "US long / DE short", "DE long / US short")
    trades["Zsig"] = zsig.loc[trades.index]
    trades.index.name = "Date"

    # Print metrics
    print(f"\n=== Strategy A (L/S)  thr {thresh}  FutZ source: {FUT_SIGNAL_SOURCE.upper()} ===")
    print(f" (hedge) used for 'hedged' z: {beta_f:.6f}  |  : {alpha_f:.6f}")
    print(f"Months: {len(df)}  |  Active months: {len(trades)}")
    print(f"Cumulative Gain: {mets['cum_gain']:.2%}")
    print(f"Max Drawdown: {mets['max_dd']:.2%}")
    print(f"Win/Loss ratio: {mets['win_loss_ratio']:.3f}  (Wins={mets['wins']}, Losses={mets['losses']})")
    print(f"VaR 95% (1-month): {mets['var95']:.2%}")
    print(f"Annualized Volatility: {mets['ann_vol']:.2%}")
    print(f"CAGR: {mets['cagr']:.2%}")
    print(f"Total Risk Ratio (CAGR/Ann.Vol): {mets['trr']:.3f}")
    print("\n--- Per-Trade P&L ---")
    print(trades[["Bias","Zsig","DE_ret_fwd","US_ret_fwd","Strat_ret"]]
          .to_string(float_format=lambda x: f"{x:,.4f}"))

    # Save CSV
    out = pd.concat(
        [
            fut[["Spread_1x1","Z_1x1","Spread_hedged","Z_hedged"]],
            df[["DE_ret_fwd","US_ret_fwd","Strat_ret"]],
            equity.rename("Equity_A"),
            drawdown.rename("Drawdown_A"),
        ], axis=1
    )
    out_path = f"/content/strategyA_LS_thr_{str(thresh).replace('.','p')}_{FUT_SIGNAL_SOURCE.lower()}.csv"
    out.to_csv(out_path)
    print(f"Saved CSV -> {out_path}")

# ---- Run Strategy A for all thresholds ----
print("Running Strategy A (L/S) for thresholds:", THRESH_LIST)
for thr in THRESH_LIST:
    run_strategy_A_pair(thr)

# ==========================================================
# STRATEGY B  Yield-only rule at 1 (unchanged, long-only)
# ==========================================================
panelB = fut.join(yld[["Z_yield"]], how="inner").join(ret_df, how="inner").dropna()

long_DE_B = (panelB["Z_yield"] <= -1.0)   # z  -1  -> long DE
long_US_B = (panelB["Z_yield"] >= +1.0)   # z  +1  -> long US

panelB["Strat_ret"] = 0.0
panelB.loc[long_DE_B, "Strat_ret"] = panelB.loc[long_DE_B, "DE_ret_fwd"]
panelB.loc[long_US_B, "Strat_ret"] = panelB.loc[long_US_B, "US_ret_fwd"]

metsB = compute_metrics(panelB["Strat_ret"])
equityB, drawdownB = metsB["equity"], metsB["drawdown"]

# Entries  Strategy B
figB1, axB1 = plt.subplots(figsize=(12,5.6))
axB1.plot(fut.index, fut["DE_fut"], label="DE futures")
axB1.plot(fut.index, fut["US_fut"], label="US futures")
de_entries_B = panelB.index[long_DE_B]; us_entries_B = panelB.index[long_US_B]
axB1.scatter(de_entries_B, fut.loc[de_entries_B, "DE_fut"], marker="^", s=90, label="ENTRY: Long DE (|Z_yield|1)", zorder=5)
axB1.scatter(us_entries_B, fut.loc[us_entries_B, "US_fut"], marker="v", s=90, label="ENTRY: Long US (|Z_yield|1)", zorder=5)
axB1.set_title("Strategy B Entries  Yield-only z rule (1)")
axB1.set_ylabel("Price"); axB1.set_xlabel("Date"); axB1.grid(True, alpha=0.3)
axB1.xaxis.set_major_formatter(DATE_FMT_M)
annotate_last(axB1, fut.index, fut["DE_fut"].values, fmt="{:.2f}")
annotate_last(axB1, fut.index, fut["US_fut"].values, fmt="{:.2f}")
axB1.legend(); finalize(figB1)

# Equity / Returns / Drawdown  Strategy B
figB2, axB2 = plt.subplots(figsize=(11,5.0))
axB2.plot(panelB.index, equityB, label="Equity")
axB2.set_title("Strategy B  Equity Curve (Yield-only 1)")
axB2.set_ylabel("Growth of $1"); axB2.set_xlabel("Date"); axB2.grid(True, alpha=0.3)
axB2.xaxis.set_major_formatter(DATE_FMT_M)
annotate_last(axB2, panelB.index, equityB.values, fmt="{:.3f}")
axB2.legend(); finalize(figB2)

figB3, axB3 = plt.subplots(figsize=(11,4.2))
colorsB = np.where(panelB["Strat_ret"] >= 0, "tab:blue", "tab:orange")
axB3.bar(panelB.index, panelB["Strat_ret"].values, width=20, align="center", color=colorsB)
axB3.axhline(0, linewidth=1)
axB3.set_title("Strategy B  Monthly Returns (Wins blue / Losses orange)")
axB3.set_ylabel("Return"); axB3.set_xlabel("Date"); axB3.grid(True, alpha=0.25)
axB3.xaxis.set_major_formatter(DATE_FMT_M)
finalize(figB3)

figB4, axB4 = plt.subplots(figsize=(11,3.8))
axB4.plot(panelB.index, drawdownB)
axB4.set_title("Strategy B  Drawdown")
axB4.set_ylabel("Drawdown"); axB4.set_xlabel("Date"); axB4.grid(True, alpha=0.3)
axB4.xaxis.set_major_formatter(DATE_FMT_M)
finalize(figB4)

# Per-trade tables + Metrics
tradesB = panelB.loc[long_DE_B | long_US_B, ["Z_yield","DE_ret_fwd","US_ret_fwd","Strat_ret"]].copy()
tradesB["Leg"] = np.where(long_DE_B.loc[tradesB.index], "DE", "US")
tradesB.index.name = "Date"

print("\n=== Strategy B  Per-Trade P&L (yield-only 1) ===")
print(tradesB[["Leg","Z_yield","DE_ret_fwd","US_ret_fwd","Strat_ret"]]
      .to_string(float_format=lambda x: f"{x:,.4f}"))

print("\n=== Strategy B  Metrics ===")
print(f"Months: {len(panelB)}  |  Active months: {len(tradesB)}")
print(f"Cumulative Gain: {metsB['cum_gain']:.2%}")
print(f"Max Drawdown: {metsB['max_dd']:.2%}")
print(f"Win/Loss ratio: {metsB['win_loss_ratio']:.3f}  (Wins={metsB['wins']}, Losses={metsB['losses']})")
print(f"VaR 95% (1-month): {metsB['var95']:.2%}")
print(f"Annualized Volatility: {metsB['ann_vol']:.2%}")
print(f"CAGR: {metsB['cagr']:.2%}")
print(f"Total Risk Ratio (CAGR / Ann.Vol): {metsB['trr']:.3f}")

# ---------------- Save outputs ----------------
outB = pd.concat(
    [
        yld[["Spread_bps","Z_yield"]],
        fut[["DE_fut","US_fut"]],
        panelB[["Z_yield","DE_ret_fwd","US_ret_fwd","Strat_ret"]],
        metsB["equity"].rename("Equity_B"),
        metsB["drawdown"].rename("Drawdown_B"),
    ], axis=1
)
outB.to_csv("/content/strategyB_yield1sigma_output.csv")

# Full OLS summaries (yield & futures)
print("\n=== Yield OLS Summary ==="); print(ols_yield.summary())
print("\n=== Futures OLS (hedge) Summary ==="); print(ols_fut.summary())




```
