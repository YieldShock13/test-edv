```python
# ============================================================
# Mag7 Backtest (Quarterly) with Two Constraint Regimes
# TRAIN_QUARTERS = 4 (start after first four quarters)
# ============================================================

import pandas as pd
import numpy as np
import glob, os
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Optional HTTP (SEC API). If not available, we fallback.
try:
    import requests
except Exception:
    requests = None

# -----------------------------
# Config
# -----------------------------
TRAIN_QUARTERS = 4            # <-- start after 4 quarters (rolling, continuous training)
ANNUALIZE      = 4            # quarterly -> annual
Z_95, Z_975    = 1.645, 1.96  # one-sided z for 95% / 97.5%
OCEAN_BG       = "#0b1e2d"
LINE_COLOR     = "#17becf"
WATERMARK      = "Orca's Capital Strategies"

# Bounds sets (two regimes)
BOUNDS_1 = (0.05, 0.20)       # 5%–20%
BOUNDS_2 = (0.10, 0.275)      # 10%–27.5%

# SEC token (DIFFERENT SOURCE from returns/RF)
SEC_API_TOKEN  = "96088fcbaf24eb3b90dab3797d56b6a1123ae775f3130709eb341fcffc13da1c"

TICKERS = ["AAPL","AMZN","GOOGL","META","MSFT","NVDA","TSLA"]

# -----------------------------
# Load returns (Investing.com monthly "Change %")
# -----------------------------
def load_monthly_returns_from_change_pct(path):
    df = pd.read_csv(path)
    if "Date" not in df.columns or "Change %" not in df.columns:
        raise ValueError(f"Need 'Date' and 'Change %' in {path}")
    df["Date"] = pd.to_datetime(df["Date"])
    r = df["Change %"].astype(str).str.replace("%","",regex=False)
    r = pd.to_numeric(r.str.replace(",","",regex=False), errors="coerce")/100.0
    s = pd.Series(r.values, index=df["Date"]).sort_index()
    return s.resample("M").last()  # keep exactly as base

def monthly_to_quarterly_returns(r_monthly):
    # geometric compounding of three months to quarter-end
    return (1.0 + r_monthly).resample("QE").prod() - 1.0

# Risk-free (annual level, percent)
def load_rf_quarterly_level(path):
    rf = pd.read_csv(path)
    dcol = [c for c in rf.columns if "date" in c.lower()][0]
    vcol = [c for c in rf.columns if c != dcol][0]
    rf[dcol] = pd.to_datetime(rf[dcol])
    rf[vcol] = pd.to_numeric(rf[vcol].replace(".", np.nan), errors="coerce")
    rf = rf.rename(columns={dcol:"Date", vcol:"DTB3"}).dropna().set_index("Date").sort_index()
    rf_m_level = rf["DTB3"]/100.0                      # decimal annual level
    rf_q_level = rf_m_level.resample("QE").last()
    return rf_q_level

# Build quarterly returns for all tickers
rets_q = {}
for tkr in TICKERS:
    files = sorted(glob.glob(f"{tkr} Historical Data*.csv"))
    if not files:
        raise FileNotFoundError(f"Missing: {tkr} Historical Data*.csv")
    r_m = load_monthly_returns_from_change_pct(files[0])
    rets_q[tkr] = monthly_to_quarterly_returns(r_m)

R = pd.DataFrame(rets_q).dropna(how="any")      # quarterly returns
rfL = load_rf_quarterly_level("DTB3.csv")       # annual level (decimal) per quarter
R, rfL = R.align(rfL, join="inner", axis=0)     # align by index

assets = list(R.columns)
n = len(assets)

# -----------------------------
# Price index from R (for market caps)
# -----------------------------
def build_price_index_quarterly(r_q):
    px = (1 + r_q.fillna(0)).cumprod()
    return px / px.iloc[0]

PX = R.apply(build_price_index_quarterly, axis=0)

# -----------------------------
# SEC API: Outstanding Shares -> quarterly shares series
# (DIFFERENT data source than returns/RF)
# -----------------------------
def fetch_outstanding_shares_secapi(ticker, token=SEC_API_TOKEN):
    """
    Returns quarterly shares outstanding series (index QE dates, column 'shares').
    If SEC endpoint unavailable, returns empty DF and we will fallback later.
    """
    if requests is None:
        return pd.DataFrame(columns=["shares"])
    url = "https://api.sec-api.io/float"
    params = {"ticker": ticker, "token": token}
    try:
        resp = requests.get(url, params=params, timeout=30)
        resp.raise_for_status()
        data = resp.json()
        rows = []
        iterable = data if isinstance(data, list) else data.get("data", [])
        for rec in iterable:
            dt = rec.get("date") or rec.get("period") or rec.get("reportDate")
            sh = rec.get("outstandingShares") or rec.get("outstanding_shares")
            if dt is None or sh is None:
                continue
            try:
                rows.append((pd.to_datetime(dt), float(sh)))
            except Exception:
                pass
        if not rows:
            return pd.DataFrame(columns=["shares"])
        df = pd.DataFrame(rows, columns=["Date","shares"]).sort_values("Date").set_index("Date")
        return df.resample("QE").last().ffill()
    except Exception as e:
        print(f"[SEC API warning] {ticker}: {e}")
        return pd.DataFrame(columns=["shares"])

def build_shares_panel_quarterly(tickers):
    panel = {}
    for t in tickers:
        dfq = fetch_outstanding_shares_secapi(t)
        if "shares" in dfq.columns and not dfq.empty:
            panel[t] = dfq["shares"].rename(t)
    if not panel:
        # No API access — return empty (we’ll use fallback)
        return pd.DataFrame(index=R.index, columns=tickers)
    SH = pd.concat(panel.values(), axis=1)
    SH = SH.reindex(R.index).ffill().bfill()    # fill across full backtest span
    return SH

SH = build_shares_panel_quarterly(assets)
if SH.isna().all().all() or SH.empty:
    # Fallback: use 1 share per ticker; weight ∝ price index (still UNBOUNDED)
    SH = pd.DataFrame(1.0, index=R.index, columns=assets)

# -----------------------------
# Rolling market-cap weights (UNBOUNDED) — SEC shares × price index
# -----------------------------
def rolling_mcap_weights_raw(PX, SH):
    idx = R.index
    rows = []
    fallback_quarters = 0
    for dt in idx:
        try:
            sh = SH.loc[dt].reindex(assets).astype(float)
        except KeyError:
            sh = pd.Series(np.nan, index=assets)
        px = PX.loc[dt].reindex(assets).astype(float)
        raw = (sh.values * px.values)
        if (not np.isfinite(raw).all()) or raw.sum() <= 0:
            raw = px.values.copy()
            if (not np.isfinite(raw).all()) or raw.sum() <= 0:
                raw = np.ones(n)
            fallback_quarters += 1
        w = raw / raw.sum()
        rows.append(pd.Series(w, index=assets, name=dt))
    if fallback_quarters > 0:
        print(f"[SEC] Fallback used in {fallback_quarters}/{len(idx)} quarters (missing/invalid shares).")
    return pd.DataFrame(rows)

W_mcap_raw = rolling_mcap_weights_raw(PX, SH)   # weights per quarter (applied next quarter)

# -----------------------------
# Risk & performance helpers
# -----------------------------
def cagr(r):    return (1+r).prod()**(ANNUALIZE/len(r)) - 1 if len(r) else np.nan
def ann_vol(r): return r.std(ddof=0)*np.sqrt(ANNUALIZE)
def max_dd(r):
    eq=(1+r.fillna(0)).cumprod(); peak=eq.cummax(); dd=eq/peak-1; return dd.min()
def var_emp(r, alpha=2.5): return np.percentile(r, alpha) if len(r) else np.nan
def win_loss(r): w=(r>0).sum(); l=(r<=0).sum(); return np.nan if l==0 else w/l
def calmar(r):
    mdd=max_dd(r)
    return np.nan if (np.isnan(mdd) or mdd>=0) else cagr(r)/abs(mdd)

def summarize(name, r_q, rf_level_annual_series):
    rf_ann_mean = rf_level_annual_series.reindex(r_q.index).mean()
    mu_q, sig_q = r_q.mean(), r_q.std(ddof=0)
    mu_ann, sig_ann = mu_q*ANNUALIZE, sig_q*np.sqrt(ANNUALIZE)
    sharpe = np.nan if sig_ann==0 else (mu_ann - rf_ann_mean)/sig_ann
    return {
        "Portfolio": name,
        "CAGR": cagr(r_q),
        "Ann Vol": sig_ann,
        "Sharpe": sharpe,
        "Max DD": max_dd(r_q),
        "VaR 97.5% (quarterly)": var_emp(r_q, 2.5),
        "Total Return": (1+r_q).prod()-1,
        "Calmar": calmar(r_q),
        "Win/Loss": win_loss(r_q),
        "Obs (quarters)": int(len(r_q))
    }

# -----------------------------
# Tail risk measures for per-asset scores (quarterly)
# -----------------------------
def mvar_vector(cov_df, w_prev, z):
    cov = cov_df.values
    sigma_p = np.sqrt(max(w_prev @ cov @ w_prev, 1e-12))
    contrib = cov @ w_prev
    mvar = z * (contrib / max(sigma_p,1e-12))
    return pd.Series(mvar, index=cov_df.columns)

def var_ordinal_vec(window_R, q):   # q=0.05 (95%) or 0.025 (97.5%)
    return -window_R.quantile(q, axis=0)

def es_ordinal_vec(window_R, q):
    # Expected Shortfall (CVaR): average of returns in the worst q tail (positive loss).
    var_cut = window_R.quantile(q, axis=0)
    es_vals=[]
    for col in window_R.columns:
        tail = window_R[window_R[col] <= var_cut[col]][col]
        es_vals.append(np.nan if tail.empty else -tail.mean())
    return pd.Series(es_vals, index=window_R.columns)

def cdar_from_path(r_series, alpha=0.95):
    # CDaR on drawdowns (positive)
    eq=(1+r_series.fillna(0)).cumprod()
    dd=eq/eq.cummax()-1.0
    qcut=dd.quantile(1-alpha)
    tail=dd[dd<=qcut]
    if len(tail)==0: return 0.0
    return -tail.mean()

# -----------------------------
# SLSQP wrappers under bounds
# -----------------------------
def slsqp_sum1_bounds_maximize(objfun, bounds):
    def minfun(w): return -objfun(w)
    cons = ({'type':'eq','fun':lambda w: w.sum()-1.0},)
    n = len(bounds)
    w0 = np.ones(n)/n
    res = minimize(minfun, w0, method="SLSQP", bounds=tuple(bounds), constraints=cons,
                   options={'ftol':1e-12,'maxiter':1000})
    if res.success: return res.x
    lo = np.array([b[0] for b in bounds]); hi = np.array([b[1] for b in bounds])
    w = np.clip(w0, lo, hi); return w / w.sum()

def slsqp_sum1_bounds_minimize(objfun, bounds):
    cons = ({'type':'eq','fun':lambda w: w.sum()-1.0},)
    n = len(bounds)
    w0 = np.ones(n)/n
    res = minimize(objfun, w0, method="SLSQP", bounds=tuple(bounds), constraints=cons,
                   options={'ftol':1e-12,'maxiter':1000})
    if res.success: return res.x
    lo = np.array([b[0] for b in bounds]); hi = np.array([b[1] for b in bounds])
    w = np.clip(w0, lo, hi); return w / w.sum()

# -----------------------------
# Optimizers (objectives)
# -----------------------------
def optimize_sharpe(mu_ex_q_vec, cov_q, bounds):
    def objective(w):
        num = mu_ex_q_vec @ w
        den = np.sqrt(max(w @ cov_q @ w, 1e-12))
        return num/den
    return slsqp_sum1_bounds_maximize(objective, bounds)

def optimize_score_linear(scores, bounds):
    s = np.asarray(scores, float)
    return slsqp_sum1_bounds_maximize(lambda w: np.dot(s,w), bounds)

def optimize_return_over_cdar(Rw, rf_per_q, alpha, bounds):
    X=Rw.values
    def objective(w):
        port_r=pd.Series(X@w,index=Rw.index)
        ex_mean=port_r.mean()-rf_per_q
        cdar = cdar_from_path(port_r, alpha=alpha)
        if cdar<=1e-10: cdar=1e-10
        return ex_mean/cdar
    return slsqp_sum1_bounds_maximize(objective, bounds)

def optimize_calmar(Rw, bounds):
    X=Rw.values
    def objective(w):
        port_r=pd.Series(X@w,index=Rw.index)
        c=cagr(port_r); dd=max_dd(port_r)
        if (dd==0) or np.isnan(dd) or (c<=0) or np.isnan(c): return 0.0
        return c/abs(dd)
    return slsqp_sum1_bounds_maximize(objective, bounds)

def optimize_min_variance(cov_q, bounds):
    return slsqp_sum1_bounds_minimize(lambda w: max(w @ cov_q @ w, 1e-16), bounds)

# -----------------------------
# Backtest engines (generic per-bounds)
# -----------------------------
def run_fixed_quarterly(target_w):
    start=TRAIN_QUARTERS
    pr=[]; W=[]
    for t in range(start, len(R)-1):
        pr.append(float(target_w @ R.iloc[t+1].values))
        W.append(pd.Series(target_w, index=assets))
    idx = R.index[start+1:]
    return pd.Series(pr, index=idx), pd.DataFrame(W, index=idx)

def run_mcap_quarterly_unbounded(W_mcap):
    start=TRAIN_QUARTERS
    pr=[]; W=[]
    for t in range(start, len(R)-1):
        w = W_mcap.iloc[t].reindex(assets).values
        w = np.where(np.isfinite(w), w, 0.0)
        if w.sum() <= 0: w = np.ones(n)/n
        else: w = w / w.sum()
        pr.append(float(w @ R.iloc[t+1].values))
        W.append(pd.Series(w, index=assets))
    idx = R.index[start+1:]
    return pd.Series(pr, index=idx), pd.DataFrame(W, index=idx)

def run_all_strategies_for_bounds(wmin, wmax, label_suffix="(bounds)"):
    bounds=[(wmin, wmax)]*n
    start=TRAIN_QUARTERS
    # Equal-weight is independent of bounds but include for reference
    port_eq, W_eq = run_fixed_quarterly(np.ones(n)/n)

    # Rolling market-cap (UNBOUNDED) — included in every table for comparison
    port_mcap, W_mcap = run_mcap_quarterly_unbounded(W_mcap_raw)

    # Containers
    results = {}
    weights = {}

    # Sharpe
    pr=[]; W=[]
    for t in range(start, len(R)-1):
        win=slice(t-TRAIN_QUARTERS, t)
        Rw=R.iloc[win]
        rf_q_level_mean=rfL.iloc[win].mean(); rf_per_q=rf_q_level_mean/ANNUALIZE
        mu_ex_q_vec=(Rw.mean().values - rf_per_q)
        cov_q=Rw.cov().values
        w=optimize_sharpe(mu_ex_q_vec, cov_q, bounds)
        pr.append(float(w @ R.iloc[t+1].values)); W.append(pd.Series(w,index=assets))
    idx=R.index[start+1:]
    results[f"Sharpe {wmin*100:.0f}–{wmax*100:.1f}%"]=pd.Series(pr,index=idx)
    weights[f"Sharpe {wmin*100:.0f}–{wmax*100:.1f}%"]=pd.DataFrame(W,index=idx)

    # ER / MVaR (95%, 97.5%)
    for z,tag in [(Z_95,"95%"),(Z_975,"97.5%")]:
        pr=[]; W=[]; w_prev=np.ones(n)/n
        for t in range(start, len(R)-1):
            win=slice(t-TRAIN_QUARTERS, t)
            Rw=R.iloc[win]
            rf_q_level_mean=rfL.iloc[win].mean(); rf_per_q=rf_q_level_mean/ANNUALIZE
            mu_ex=(Rw.mean()-rf_per_q)
            mvar=mvar_vector(Rw.cov(), w_prev, z).replace([np.inf,-np.inf],np.nan).abs()
            mvar=mvar.where(mvar>0, np.nan)
            scores=(mu_ex/mvar).replace([np.inf,-np.inf],np.nan).fillna(0.0).values
            w=optimize_score_linear(scores, bounds)
            pr.append(float(w @ R.iloc[t+1].values)); W.append(pd.Series(w,index=assets))
            w_prev=w
        results[f"ER/MVaR {tag} {wmin*100:.0f}–{wmax*100:.1f}%"]=pd.Series(pr,index=idx)
        weights[f"ER/MVaR {tag} {wmin*100:.0f}–{wmax*100:.1f}%"]=pd.DataFrame(W,index=idx)

    # ER / VaR (95%, 97.5%)
    for q,tag in [(0.05,"95%"),(0.025,"97.5%")]:
        pr=[]; W=[]
        for t in range(start, len(R)-1):
            win=slice(t-TRAIN_QUARTERS, t)
            Rw=R.iloc[win]
            rf_q_level_mean=rfL.iloc[win].mean(); rf_per_q=rf_q_level_mean/ANNUALIZE
            mu_ex=(Rw.mean()-rf_per_q)
            var_i=var_ordinal_vec(Rw,q).replace([np.inf,-np.inf],np.nan)
            var_i=var_i.where(var_i>0,np.nan)
            scores=(mu_ex/var_i).replace([np.inf,-np.inf],np.nan).fillna(0.0).values
            w=optimize_score_linear(scores, bounds)
            pr.append(float(w @ R.iloc[t+1].values)); W.append(pd.Series(w,index=assets))
        results[f"ER/VaR {tag} {wmin*100:.0f}–{wmax*100:.1f}%"]=pd.Series(pr,index=idx)
        weights[f"ER/VaR {tag} {wmin*100:.0f}–{wmax*100:.1f}%"]=pd.DataFrame(W,index=idx)

    # ER / ES (95%, 97.5%)
    for q,tag in [(0.05,"95%"),(0.025,"97.5%")]:
        pr=[]; W=[]
        for t in range(start, len(R)-1):
            win=slice(t-TRAIN_QUARTERS, t)
            Rw=R.iloc[win]
            rf_q_level_mean=rfL.iloc[win].mean(); rf_per_q=rf_q_level_mean/ANNUALIZE
            mu_ex=(Rw.mean()-rf_per_q)
            es_i=es_ordinal_vec(Rw,q).replace([np.inf,-np.inf],np.nan)
            es_i=es_i.where(es_i>0,np.nan)
            scores=(mu_ex/es_i).replace([np.inf,-np.inf],np.nan).fillna(0.0).values
            w=optimize_score_linear(scores, bounds)
            pr.append(float(w @ R.iloc[t+1].values)); W.append(pd.Series(w,index=assets))
        results[f"ER/ES {tag} {wmin*100:.0f}–{wmax*100:.1f}%"]=pd.Series(pr,index=idx)
        weights[f"ER/ES {tag} {wmin*100:.0f}–{wmax*100:.1f}%"]=pd.DataFrame(W,index=idx)

    # Return / CDaR (95%, 97.5%)
    for alpha,tag in [(0.95,"95%"),(0.975,"97.5%")]:
        pr=[]; W=[]
        for t in range(start, len(R)-1):
            win=slice(t-TRAIN_QUARTERS, t)
            Rw=R.iloc[win]
            rf_q_level_mean=rfL.iloc[win].mean(); rf_per_q=rf_q_level_mean/ANNUALIZE
            w=optimize_return_over_cdar(Rw, rf_per_q, alpha, bounds)
            pr.append(float(w @ R.iloc[t+1].values)); W.append(pd.Series(w,index=assets))
        results[f"Return/CDaR {tag} {wmin*100:.0f}–{wmax*100:.1f}%"]=pd.Series(pr,index=idx)
        weights[f"Return/CDaR {tag} {wmin*100:.0f}–{wmax*100:.1f}%"]=pd.DataFrame(W,index=idx)

    # Calmar
    pr=[]; W=[]
    for t in range(start, len(R)-1):
        win=slice(t-TRAIN_QUARTERS, t)
        Rw=R.iloc[win]
        w=optimize_calmar(Rw, bounds)
        pr.append(float(w @ R.iloc[t+1].values)); W.append(pd.Series(w,index=assets))
    results[f"Calmar {wmin*100:.0f}–{wmax*100:.1f}%"]=pd.Series(pr,index=idx)
    weights[f"Calmar {wmin*100:.0f}–{wmax*100:.1f}%"]=pd.DataFrame(W,index=idx)

    # Minimum Variance
    pr=[]; W=[]
    for t in range(start, len(R)-1):
        win=slice(t-TRAIN_QUARTERS, t)
        Rw=R.iloc[win]
        cov_q=Rw.cov().values
        w=optimize_min_variance(cov_q, bounds)
        pr.append(float(w @ R.iloc[t+1].values)); W.append(pd.Series(w,index=assets))
    results[f"MinVar {wmin*100:.0f}–{wmax*100:.1f}%"]=pd.Series(pr,index=idx)
    weights[f"MinVar {wmin*100:.0f}–{wmax*100:.1f}%"]=pd.DataFrame(W,index=idx)

    # Add equal & rolling mcap (unbounded) to the dicts for reporting convenience
    results["Equal-Weight"] = port_eq
    weights["Equal-Weight"] = W_eq
    results["Rolling Market-Cap (UNBOUNDED)"] = port_mcap
    weights["Rolling Market-Cap (UNBOUNDED)"] = W_mcap

    return results, weights

# -----------------------------
# Run both bounds regimes
# -----------------------------
res_520, W_520     = run_all_strategies_for_bounds(0.05, 0.20)
res_10275, W_10275 = run_all_strategies_for_bounds(0.10, 0.275)

# -----------------------------
# Summaries (two separate tables)
# -----------------------------
def fmt_pct(x):   return "—" if pd.isna(x) else f"{x*100:.2f}%"
def fmt_ratio(x): return "—" if pd.isna(x) else f"{x:.2f}"
def fmt_var(x):   return "—" if pd.isna(x) else f"{x*100:.2f}%"

def make_summary(res_dict, title):
    rows=[]
    for name, r in res_dict.items():
        rows.append(summarize(name, r, rfL))
    df = pd.DataFrame(rows)
    out = df.copy()
    out["CAGR"]                 = out["CAGR"].apply(fmt_pct)
    out["Ann Vol"]              = out["Ann Vol"].apply(fmt_pct)
    out["Sharpe"]               = out["Sharpe"].round(2)
    out["Max DD"]               = out["Max DD"].apply(fmt_pct)
    out["VaR 97.5% (quarterly)"]= out["VaR 97.5% (quarterly)"].apply(fmt_var)
    out["Total Return"]         = out["Total Return"].apply(fmt_pct)
    out["Calmar"]               = out["Calmar"].apply(fmt_ratio)
    out["Win/Loss"]             = out["Win/Loss"].apply(fmt_ratio)
    print(f"\n=== {title} ===")
    print(out[["Portfolio","CAGR","Ann Vol","Sharpe","Max DD",
               "VaR 97.5% (quarterly)","Total Return","Calmar","Win/Loss","Obs (quarters)"]]
          .to_string(index=False))
    return df, out

summary_raw_520, summary_520 = make_summary(res_520, "Summary — Bounds 5%–20% (TRAIN_QUARTERS=4)")
summary_raw_10275, summary_10275 = make_summary(res_10275, "Summary — Bounds 10%–27.5% (TRAIN_QUARTERS=4)")

# -----------------------------
# Diagnostics for Rolling Market-Cap
# -----------------------------
W_mcap = W_520["Rolling Market-Cap (UNBOUNDED)"]  # same under both regimes
print("\n=== Average Market-Cap Weights (UNBOUNDED, over OOS) ===")
print(W_mcap.mean().round(4).to_string())
print("\n=== Rolling Market-Cap Weights — first few OOS quarters ===")
print(W_mcap.head(12).round(4).to_string())
print("\n=== Rolling Market-Cap Weights — last few OOS quarters ===")
print(W_mcap.tail(12).round(4).to_string())

# -----------------------------
# Plots (equity curves + stacked weights)
# -----------------------------
def ocean_axes(ax, title):
    ax.set_facecolor(OCEAN_BG); ax.figure.set_facecolor(OCEAN_BG)
    ax.set_title(title, color="white", fontsize=15, fontweight="bold", pad=18)
    ax.tick_params(colors="white")
    ax.spines["bottom"].set_color("white"); ax.spines["left"].set_color("white")
    ax.spines["top"].set_visible(False);   ax.spines["right"].set_visible(False)
    ax.grid(alpha=0.2, color="white")

def add_watermark(ax):
    ax.text(0.5, 0.05, WATERMARK, transform=ax.transAxes,
            fontsize=16, color="white", alpha=0.15,
            ha="center", va="center", rotation=30)

def equity_curve(r, title):
    eq=(1+r.fillna(0)).cumprod()
    fig, ax = plt.subplots(figsize=(12,6))
    ax.plot(eq.index, eq.values, color=LINE_COLOR, linewidth=1.8, label=title)
    ocean_axes(ax, f"Equity Curve — {title}"); add_watermark(ax)
    ax.set_ylabel("Growth of $1", color="white")
    last_date, last_val = eq.index[-1], eq.iloc[-1]
    ax.scatter([last_date],[last_val], color="red")
    ax.annotate(f"{last_val:.2f}x", xy=(last_date,last_val), xytext=(8,8),
                textcoords="offset points", fontsize=10, color="white", weight="bold")
    ax.legend(facecolor=OCEAN_BG, edgecolor="none", labelcolor="white")
    plt.tight_layout(); plt.show()

def stacked_weights(W, title):
    fig, ax = plt.subplots(figsize=(12,6))
    W.plot(kind="area", ax=ax, linewidth=0)
    ocean_axes(ax, title); add_watermark(ax)
    ax.set_ylabel("Weight", color="white"); ax.set_ylim(0,1)
    plt.tight_layout(); plt.show()

# --- Equity curves: key strategies for both regimes + mcap unbounded
for name in [
    "Equal-Weight",
    "Rolling Market-Cap (UNBOUNDED)",
    "Sharpe 5–20.0%",
    "Sharpe 10–27.5%",
    "ER/MVaR 95% 5–20.0%",
    "ER/MVaR 95% 10–27.5%",
    "ER/VaR 95% 5–20.0%",
    "ER/VaR 95% 10–27.5%",
    "ER/ES 95% 5–20.0%",
    "ER/ES 95% 10–27.5%",
    "Return/CDaR 95% 5–20.0%",
    "Return/CDaR 95% 10–27.5%",
    "Calmar 5–20.0%",
    "Calmar 10–27.5%",
    "MinVar 5–20.0%",
    "MinVar 10–27.5%",
]:
    # pick from the right dict
    series = res_520.get(name) if name in res_520 else res_10275.get(name)
    if series is None:
        # try alternate because the name includes formatted bounds
        series = res_10275.get(name, None)
    if series is not None:
        equity_curve(series, name)

# --- Stacked weights: a subset (to avoid wall-of-plots); add more if you like
for title, Wset in [
    ("Weights — Rolling Market-Cap (UNBOUNDED)", W_mcap),
    ("Weights — Sharpe 5–20%",  W_520["Sharpe 5–20.0%"]),
    ("Weights — Sharpe 10–27.5%", W_10275["Sharpe 10–27.5%"]),
    ("Weights — ER/MVaR 95% 5–20%",  W_520["ER/MVaR 95% 5–20.0%"]),
    ("Weights — ER/MVaR 95% 10–27.5%", W_10275["ER/MVaR 95% 10–27.5%"]),
    ("Weights — MinVar 5–20%",  W_520["MinVar 5–20.0%"]),
    ("Weights — MinVar 10–27.5%", W_10275["MinVar 10–27.5%"]),
]:
    stacked_weights(Wset, title)




```
