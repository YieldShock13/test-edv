```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# =====================================================
# SETTINGS
# =====================================================

START_DATE = "2020-02-09"
LOOKBACK = 52
USDZAR_SPOT_T0 = 16.04

ZAR_CASH = 30000
USD_CASH = 300

ZAR_RATE = 0.06 / 52
USD_RATE = 0.03 / 52

TRUW_DIV = 0.085 / 52
IEF_COUPON = 0.04 / 52

RF_ANNUAL = 0.045
RF_WEEKLY = (1 + RF_ANNUAL)**(1/52) - 1

VAR_LEVELS = [0.95, 0.99]

# =====================================================
# LOADERS
# =====================================================

def load_price_series(filename):
    df = pd.read_csv(filename)
    df["Date"] = pd.to_datetime(df["Date"])
    df = df.sort_values("Date")
    df["Price"] = df["Price"].astype(str).str.replace(",", "", regex=False).astype(float)
    df = df[["Date","Price"]].set_index("Date")
    df = df.resample("W-FRI").last()
    df["ret"] = df["Price"].pct_change(fill_method=None)
    return df["ret"]

def load_monthly_returns(filename):
    df = pd.read_csv(filename)
    df["Date"] = pd.to_datetime(df["Date"])
    df = df.sort_values("Date")
    df["Change %"] = df["Change %"].astype(str).str.replace("%","").astype(float)/100
    monthly = df[["Date","Change %"]].set_index("Date")["Change %"]
    weekly_index = pd.date_range(monthly.index.min(), monthly.index.max(), freq="W-FRI")
    weekly = pd.Series(0.0, index=weekly_index)
    for d,r in monthly.items():
        eligible = weekly.index[weekly.index <= d]
        if len(eligible)>0:
            weekly.loc[eligible[-1]] = r
    return weekly

# =====================================================
# LOAD DATA
# =====================================================

truw      = load_price_series("TRUJ Historical Data (2).csv") + TRUW_DIV
newgold   = load_price_series("GLDJ Historical Data (2).csv")
msci_zar  = load_price_series("0P0000ZSS4 Historical Data.csv")
msci_em   = load_price_series("STXEMGJ Historical Data (1).csv")
albi      = load_price_series("0P00015WR7 Historical Data.csv")

urth      = load_price_series("URTH Historical Data.csv")
nflx      = load_price_series("NFLX Historical Data.csv")
btc       = load_price_series("Bitcoin Historical Data (1).csv")
ief       = load_price_series("IEF Historical Data (6).csv") + IEF_COUPON

usdzar    = load_price_series("USD_ZAR Historical Data (5).csv")
eurusd    = load_price_series("EUR_USD Historical Data (1).csv")

pere1     = load_price_series("0P0001K7XM Historical Data (2).csv")
pere2     = load_price_series("0P0001K7XN Historical Data.csv")
laurium   = load_monthly_returns("Laurium_Market_Neutral_Monthly_Returns_Investing_Format.csv")

data = pd.concat(
    [truw,newgold,msci_zar,msci_em,albi,
     urth,nflx,btc,ief,
     usdzar,eurusd,
     pere1,pere2,laurium],
    axis=1
)

data.columns = [
    "TRUW","NEWGOLD","MSCI_ZAR","MSCI_EM","ALBI",
    "URTH","NFLX","BTC","IEF",
    "USDZAR","EURUSD",
    "PERE1","PERE2","LAURIUM"
]

data = data.loc[START_DATE:].dropna()

# =====================================================
# FX HELPERS
# =====================================================

def usd_to_zar(r):
    return (1+r)*(1+data["USDZAR"])-1

def zar_to_usd(r):
    return (1+r)/(1+data["USDZAR"])-1

# =====================================================
# BUILD RETURNS
# =====================================================

def build_returns(base="ZAR"):

    R = pd.DataFrame(index=data.index)
    ief_eur = (1 + data["IEF"]) / (1 + data["EURUSD"]) - 1

    if base=="ZAR":
        R["TRUW"]=data["TRUW"]
        R["NEWGOLD"]=data["NEWGOLD"]
        R["MSCI_ZAR"]=data["MSCI_ZAR"]
        R["MSCI_EM"]=data["MSCI_EM"]
        R["ALBI"]=data["ALBI"]
        R["PERE1"]=data["PERE1"]
        R["PERE2"]=data["PERE2"]
        R["LAURIUM"]=data["LAURIUM"]

        R["URTH"]=usd_to_zar(data["URTH"])
        R["NFLX"]=usd_to_zar(data["NFLX"])
        R["BTC"]=usd_to_zar(data["BTC"])

        eurzar=(1+data["EURUSD"])*(1+data["USDZAR"])-1
        R["IEF_EUR"]=(1+ief_eur)*(1+eurzar)-1

        R["ZAR_CASH"]=ZAR_RATE
        R["USD_CASH"]=usd_to_zar(USD_RATE)

    if base=="USD":
        R["TRUW"]=zar_to_usd(data["TRUW"])
        R["NEWGOLD"]=zar_to_usd(data["NEWGOLD"])
        R["MSCI_ZAR"]=zar_to_usd(data["MSCI_ZAR"])
        R["MSCI_EM"]=zar_to_usd(data["MSCI_EM"])
        R["ALBI"]=zar_to_usd(data["ALBI"])
        R["PERE1"]=zar_to_usd(data["PERE1"])
        R["PERE2"]=zar_to_usd(data["PERE2"])
        R["LAURIUM"]=zar_to_usd(data["LAURIUM"])

        R["URTH"]=data["URTH"]
        R["NFLX"]=data["NFLX"]
        R["BTC"]=data["BTC"]

        R["IEF_EUR"]=(1+ief_eur)*(1+data["EURUSD"])-1

        R["USD_CASH"]=USD_RATE
        R["ZAR_CASH"]=zar_to_usd(ZAR_RATE)

    return R

# =====================================================
# FROZEN NAV
# =====================================================

nav_zar = {
    "TRUW":15779,"NEWGOLD":21977,"MSCI_ZAR":44490,"MSCI_EM":26552,
    "ALBI":29332,"PERE1":25000,"PERE2":25000,"LAURIUM":10000,
    "URTH":1500*USDZAR_SPOT_T0,
    "IEF_EUR":3200*USDZAR_SPOT_T0,
    "BTC":1234*USDZAR_SPOT_T0,
    "NFLX":6191,
    "ZAR_CASH":ZAR_CASH,
    "USD_CASH":USD_CASH*USDZAR_SPOT_T0
}

# =====================================================
# CORE RUN
# =====================================================

def run(base="ZAR", include_btc=True):

    print("\n==============================")
    print("Base:",base," BTC:",include_btc)
    print("==============================")

    R=build_returns(base)
    nav=pd.Series(nav_zar)

    if base=="USD":
        nav=nav/USDZAR_SPOT_T0

    if not include_btc:
        nav["BTC"]=0

    w=nav.reindex(R.columns,fill_value=0)
    w=w/w.sum()

    print("\nPosition Sizes:")
    print(nav)

    print("\nWeights:")
    print(w.sort_values(ascending=False))

    port=(R*w).sum(axis=1)

    # Basic Metrics
    cagr=(1+port).prod()**(52/len(port))-1
    vol=port.std()*np.sqrt(52)
    equity=(1+port).cumprod()
    dd=equity/equity.cummax()-1
    maxdd=dd.min()

    # VaR
    vars={lvl:-np.quantile(port,1-lvl) for lvl in VAR_LEVELS}

    # Regression vs URTH
    bench=R["URTH"]
    df=pd.concat([port,bench],axis=1).dropna()
    X=np.column_stack([np.ones(len(df)),df.iloc[:,1]])
    beta_hat=np.linalg.lstsq(X,df.iloc[:,0],rcond=None)[0]
    alpha=beta_hat[0]*52
    beta=beta_hat[1]

    # Sharpe
    excess=port-RF_WEEKLY
    sharpe=(excess.mean()*52)/(port.std()*np.sqrt(52))

    # Calmar
    calmar=cagr/abs(maxdd)

    # Information Ratio
    active=port-bench
    tracking=active.std()*np.sqrt(52)
    ir=(active.mean()*52)/tracking

    # Conditional Beta
    down=bench<0
    cond_beta=np.cov(port[down],bench[down])[0,1]/np.var(bench[down])

    # Risk Contribution
    cov=R.cov()
    wv=w.values
    port_var=wv.T@cov.values@wv
    port_vol=np.sqrt(port_var)
    marginal=cov.values@wv/port_vol
    rc=pd.Series(wv*marginal/port_vol,index=w.index).sort_values(ascending=False)

    # Print
    print("\nCAGR:",cagr)
    print("Vol:",vol)
    print("MaxDD:",maxdd)
    print("Sharpe:",sharpe)
    print("Calmar:",calmar)
    print("Alpha:",alpha)
    print("Beta:",beta)
    print("Information Ratio:",ir)
    print("Conditional Beta:",cond_beta)
    print("VaR:",vars)
    print("\nRisk Contribution:")
    print(rc)

    # Charts
    plt.figure(); equity.plot(title="Equity"); plt.show()
    plt.figure(); dd.plot(title="Drawdown"); plt.show()
    plt.figure(); (port.rolling(LOOKBACK).std()*np.sqrt(52)).plot(title="Rolling Vol"); plt.show()
    plt.figure(); plt.scatter(bench,port); plt.title("Portfolio vs URTH"); plt.show()

# =====================================================
# RUN ALL
# =====================================================

run("ZAR",True)
run("ZAR",False)
run("USD",True)
run("USD",False)







```
