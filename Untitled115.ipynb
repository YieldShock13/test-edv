```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# =========================
# CONFIG
# =========================

START_DATE      = "2005-01-01"
END_DATE        = "2025-11-30"
BACKFILL_START  = "2008-11-14"   # regression / analysis start after backfill
FREQ            = "W-FRI"        # weekly on Fridays
PER_YR          = 52             # periods per year for weekly data

# --- File mapping: ticker -> CSV in current working dir
FILE_MAP = {
    # Portfolio names
    "AGLJ":  "AGLJ Historical Data.csv",
    "ABI":   "ABI Historical Data.csv",
    "BHPJ":  "BHP Historical Data.csv",          # use ADR or local, as provided
    "BIDJ":  "BIDJ Historical Data.csv",
    "BTIJ":  "BTIJ Historical Data.csv",
    "BVTJ":  "BVTJ Historical Data.csv",
    "CFRJ":  "CFRJ Historical Data.csv",
    "EMHJ":  "EMHJ Historical Data.csv",
    "REMJ":  "REMJ Historical Data.csv",
    "RNIJ":  "RNIJ Historical Data.csv",
    "SHPJ":  "SHPJ Historical Data.csv",
    "TBSJ":  "TBSJ Historical Data.csv",
    "VALJ":  "VALJ Historical Data (1).csv",
    # Benchmarks
    "JALSH":  "FTSE_JSE All Share Historical Data (1).csv",
    "JTOP40": "South Africa Top 40 Historical Data (1).csv",
}

PORT_TICKERS  = ["AGLJ","ABI","BHPJ","BIDJ","BTIJ","BVTJ","CFRJ",
                 "EMHJ","REMJ","RNIJ","SHPJ","TBSJ","VALJ"]
BENCH_TICKERS = ["JALSH","JTOP40"]

# --- Current mark-to-market weights from statement (% of book, in decimals)
W_MTM = pd.Series({
    "AGLJ": 0.1071,   # Anglo American
    "ABI":  0.0282,   # Anheuser-Busch InBev
    "BHPJ": 0.0186,   # BHP
    "BIDJ": 0.0820,   # Bidcorp
    "BTIJ": 0.1328,   # British American Tobacco
    "BVTJ": 0.0764,   # Bidvest
    "CFRJ": 0.1688,   # Richemont
    "EMHJ": 0.0003,   # eMedia (tiny)
    "REMJ": 0.0698,   # Remgro
    "RNIJ": 0.1239,   # Reinet
    "SHPJ": 0.0918,   # Shoprite
    "TBSJ": 0.0721,   # Tiger Brands
    "VALJ": 0.0199,   # Valterra Platinum
})
W_MTM = W_MTM / W_MTM.sum()   # force to 1.0

# --- Cost-based weights: FROM STATEMENT (Total Cost column)
COST_VALUE_ZAR = {
    "AGLJ": 451_442.00,
    "ABI":  467_957.00,
    "BHPJ":      0.00,   # if cost is zero in statement, leave 0
    "BIDJ": 653_829.00,
    "BTIJ": 729_085.00,
    "BVTJ": 760_454.00,
    "CFRJ": 735_033.00,
    "EMHJ":   2_764.00,
    "REMJ": 481_854.00,
    "RNIJ": 623_667.00,
    "SHPJ":1_190_599.00,
    "TBSJ": 729_489.00,
    "VALJ": 172_612.00,
}
W_COST = pd.Series(COST_VALUE_ZAR, dtype="float64")
W_COST = W_COST / W_COST.sum()

# --- Dividend yield assumptions (simple constant annual yields, decimals).
DIV_YIELD_ANNUAL = {
    "AGLJ":  0.07,   # Anglo American
    "ABI":   0.025,  # ABI
    "BHPJ":  0.065,  # BHP
    "BIDJ":  0.024,  # Bidcorp
    "BTIJ":  0.075,  # BTI
    "BVTJ":  0.033,  # Bidvest
    "CFRJ":  0.022,  # Richemont
    "EMHJ":  0.09,   # eMedia (capped)
    "REMJ":  0.03,   # Remgro
    "RNIJ":  0.014,  # Reinet
    "SHPJ":  0.03,   # Shoprite
    "TBSJ":  0.05,   # Tiger Brands
    "VALJ":  0.08,   # Valterra
    "JALSH": 0.035,  # ALSI long-run mid-3s
    "JTOP40":0.03,   # Top40 ~3%
}

# --- Risk-free rate assumption (annualised, decimal).
RF_ANN    = 0.07
RF_WEEKLY = (1 + RF_ANN)**(1 / PER_YR) - 1

# =========================
# DATA HELPERS
# =========================

def load_investing_csv(filename):
    """Load an Investing.com-style CSV and return a daily price series."""
    df = pd.read_csv(filename)
    df.columns = [c.strip() for c in df.columns]
    if "Date" not in df.columns:
        raise ValueError(f"'Date' column not found in {filename}")
    df["Date"] = pd.to_datetime(df["Date"], dayfirst=True, errors="coerce")
    df = df.dropna(subset=["Date"]).sort_values("Date").set_index("Date")

    price_col = None
    for c in ["Price", "Close", "Last", "Close*"]:
        if c in df.columns:
            price_col = c
            break
    if price_col is None:
        raise ValueError(f"No price-like column in {filename}: {df.columns.tolist()}")

    s = pd.to_numeric(
        df[price_col].astype(str).str.replace(",", ""),
        errors="coerce"
    ).dropna()
    s.name = os.path.splitext(os.path.basename(filename))[0]
    return s

def to_weekly_returns(price, freq=FREQ):
    weekly = price.resample(freq).last().dropna()
    return weekly.pct_change().dropna()

def clean_stock_splits(r, jump_thresh=1.0, window=4):
    """
    Handle genuine stock / reverse splits:
    - Only adjust |return| >= jump_thresh (e.g. 100%).
    - Replace with median of neighbouring window.
    Returns cleaned series and index of adjusted dates.
    """
    r = r.copy()
    large = r.index[r.abs() >= jump_thresh]
    adjusted_idx = []
    for dt in large:
        loc = r.index.get_loc(dt)
        lo  = max(0, loc - window)
        hi  = min(len(r), loc + window + 1)
        neigh = r.iloc[lo:hi].drop(index=dt)
        if len(neigh) == 0:
            continue
        r.iloc[loc] = neigh.median()
        adjusted_idx.append(dt)
    flags = pd.Index(adjusted_idx)
    return r, flags

def add_dividends(weekly_ret, ticker):
    """Convert annual dividend yield into a flat weekly increment."""
    y_ann = DIV_YIELD_ANNUAL.get(ticker, 0.0)
    if y_ann == 0.0:
        return weekly_ret
    div_w = (1 + y_ann)**(1 / PER_YR) - 1
    return weekly_ret + div_w

# =========================
# LOAD ALL SERIES
# =========================

print("Files in working dir:", os.listdir())

weekly_price_rets = {}
outlier_flags = {}

for ticker, fname in FILE_MAP.items():
    if not os.path.exists(fname):
        print(f"WARNING: file for {ticker} not found: {fname}")
        continue

    s_price = load_investing_csv(fname)
    s_ret   = to_weekly_returns(s_price)
    s_ret   = s_ret.loc[START_DATE:END_DATE]

    if ticker in BENCH_TICKERS:
        # Do NOT winsorise benchmarks; keep their full history.
        s_clean = s_ret
        flags   = pd.Index([])
    else:
        s_clean, flags = clean_stock_splits(s_ret, jump_thresh=1.0, window=4)

    weekly_price_rets[ticker] = s_clean
    outlier_flags[ticker]     = flags

# Build union panel (no dropna yet)
price_panel_raw = pd.DataFrame(weekly_price_rets).sort_index()

print("\nInitial regression start dates (before backfill):")
for t in price_panel_raw.columns:
    print(f"{t}: {price_panel_raw[t].first_valid_index()}")

# =========================
# BACKFILL BIDJ BY REGRESSION TO 2008-11-14
# =========================

backfill_start_ts = pd.to_datetime(BACKFILL_START)

target = "BIDJ"
if target in price_panel_raw.columns:
    natural_start = price_panel_raw[target].first_valid_index()
    print(f"\nNatural start for {target}: {natural_start}")
    if natural_start is not None and natural_start > backfill_start_ts:
        predictors = [c for c in price_panel_raw.columns if c != target]

        df = price_panel_raw[[target] + predictors].copy()

        # Fit OLS on overlapping sample where target & predictors all non-null
        mask_fit = df[target].notna() & df[predictors].notna().all(axis=1)
        X_fit = df.loc[mask_fit, predictors].values
        y_fit = df.loc[mask_fit, target].values

        if X_fit.shape[0] >= len(predictors) + 2:
            X_design = np.column_stack([np.ones(len(X_fit)), X_fit])
            beta_hat, *_ = np.linalg.lstsq(X_design, y_fit, rcond=None)

            # Build pre-history from BACKFILL_START up to just before natural_start
            mask_pre = (df.index >= backfill_start_ts) & (df.index < natural_start)
            X_pre = df.loc[mask_pre, predictors].dropna()
            if len(X_pre) > 0:
                X_pre_design = np.column_stack([np.ones(len(X_pre)), X_pre.values])
                y_hat = X_pre_design @ beta_hat
                price_panel_raw.loc[X_pre.index, target] = y_hat
                print(f"Backfilled {target} over {len(X_pre)} weekly observations starting {X_pre.index[0]}")
        else:
            print(f"Not enough overlapping data to backfill {target}; leaving as-is.")

# =========================
# FINAL ALIGNED PANELS (PRICE & TOTAL)
# =========================

# Restrict to backfill start and end, then drop rows with any NaNs
price_panel = price_panel_raw.loc[backfill_start_ts:END_DATE].dropna(how="any")

print("\nAligned sample:")
print("Start:", price_panel.index[0], "End:", price_panel.index[-1])
print("Obs:", len(price_panel))

print("\nRegression start dates (weekly, after backfill & NaN drop):")
for t in price_panel.columns:
    print(f"{t}: {price_panel[t].first_valid_index()}")

# Build total-return panel by adding synthetic dividends
total_panel = pd.DataFrame(index=price_panel.index, columns=price_panel.columns, dtype="float64")
for t in price_panel.columns:
    total_panel[t] = add_dividends(price_panel[t], t)

# =========================
# PORTFOLIO CONSTRUCTION
# =========================

# Restrict weights to the tickers that survived / are in panel
PORT_TICKERS_IN_PANEL = [t for t in PORT_TICKERS if t in price_panel.columns]

price_port = price_panel[PORT_TICKERS_IN_PANEL]
total_port = total_panel[PORT_TICKERS_IN_PANEL]

W_MTM_use  = W_MTM.reindex(PORT_TICKERS_IN_PANEL).fillna(0.0)
W_MTM_use  = W_MTM_use / W_MTM_use.sum()

W_COST_use = W_COST.reindex(PORT_TICKERS_IN_PANEL).fillna(0.0)
W_COST_use = W_COST_use / W_COST_use.sum()

print("\n== PORTFOLIO WEIGHTS USED (MTM, % of equity) ==")
print((W_MTM_use * 100).round(2))

print("\n== PORTFOLIO WEIGHTS USED (COST-BASED, % of equity) ==")
print((W_COST_use * 100).round(2))

def portfolio_return(ret_df, weights: pd.Series):
    return (ret_df * weights.values).sum(axis=1)

# Price-only portfolios
rp_mtm_price  = portfolio_return(price_port, W_MTM_use)
rp_cost_price = portfolio_return(price_port, W_COST_use)

# Total-return portfolios
rp_mtm_total  = portfolio_return(total_port, W_MTM_use)
rp_cost_total = portfolio_return(total_port, W_COST_use)

# Benchmarks
rb_jalsh_price = price_panel["JALSH"]
rb_top40_price = price_panel["JTOP40"]
rb_jalsh_total = total_panel["JALSH"]
rb_top40_total = total_panel["JTOP40"]

# =========================
# RISK METRICS
# =========================

def cagr(r):
    if len(r) == 0:
        return np.nan
    cum = (1 + r).prod()
    yrs = len(r) / PER_YR
    return cum**(1/yrs) - 1

def ann_vol(r):
    return r.std(ddof=1) * np.sqrt(PER_YR)

def sharpe_ratio(r, rf_weekly=RF_WEEKLY):
    ex = r - rf_weekly
    denom = r.std(ddof=1)
    if denom == 0:
        return np.nan
    return (ex.mean() * PER_YR) / (denom * np.sqrt(PER_YR))

def max_drawdown(r):
    curve = (1 + r).cumprod()
    peak  = curve.cummax()
    dd    = (curve / peak) - 1.0
    return dd.min()

def hist_var(r, alpha=0.05):
    return np.quantile(r, alpha)

def capm_stats(r_p, r_m, rf_weekly=RF_WEEKLY):
    ex_p = r_p - rf_weekly
    ex_m = r_m - rf_weekly
    cov  = np.cov(ex_p, ex_m)[0, 1]
    varm = np.var(ex_m)
    if varm <= 0:
        beta = np.nan
    else:
        beta = cov / varm
    alpha_weekly = ex_p.mean() - beta * ex_m.mean() if not np.isnan(beta) else np.nan
    alpha_ann    = (1 + alpha_weekly)**PER_YR - 1 if not np.isnan(alpha_weekly) else np.nan
    te           = ann_vol(ex_p - ex_m)
    return beta, alpha_ann, te

def summarise(r, name, bench=None):
    d = {
        "CAGR":   cagr(r),
        "AnnVol": ann_vol(r),
        "Sharpe": sharpe_ratio(r),
        "MaxDD":  max_drawdown(r),
        "VaR_5pc":hist_var(r, 0.05),
    }
    if bench is not None:
        beta, alpha_ann, te = capm_stats(r, bench)
        d["Beta_vs_bench"] = beta
        d["Alpha_ann"]     = alpha_ann
        d["TE"]            = te
    return pd.Series(d, name=name)

summaries = []

# price-only vs ALSI
summaries.append(summarise(rp_mtm_price,  "Port_MTM_price_vs_ALSI",  bench=rb_jalsh_price))
summaries.append(summarise(rp_cost_price, "Port_COST_price_vs_ALSI", bench=rb_jalsh_price))
summaries.append(summarise(rb_jalsh_price,"ALSI_price",              bench=rb_jalsh_price))
summaries.append(summarise(rb_top40_price,"Top40_price",             bench=rb_jalsh_price))

# total-return vs ALSI total
summaries.append(summarise(rp_mtm_total,  "Port_MTM_total_vs_ALSI",  bench=rb_jalsh_total))
summaries.append(summarise(rp_cost_total, "Port_COST_total_vs_ALSI", bench=rb_jalsh_total))
summaries.append(summarise(rb_jalsh_total,"ALSI_total",              bench=rb_jalsh_total))
summaries.append(summarise(rb_top40_total,"Top40_total",             bench=rb_jalsh_total))

summary_table = pd.DataFrame(summaries)

# Convert selected columns to % for readability
pct_cols = ["CAGR","AnnVol","MaxDD","VaR_5pc","Alpha_ann","TE"]
for c in pct_cols:
    if c in summary_table.columns:
        summary_table[c] = summary_table[c] * 100

print("\nRisk / performance summary (weekly data, annualised; % except Sharpe/Beta):")
print(summary_table.round(3))

# =========================
# ROLLING AVERAGE CORRELATION OF CONSTITUENTS
# =========================

def rolling_avg_corr(ret_df: pd.DataFrame, window: int = 52) -> pd.Series:
    """
    Rolling average pairwise correlation of all columns in ret_df.
    Implemented manually (no rolling.apply), so no Series.corr() error.
    """
    ret_df = ret_df.dropna(how="any")
    idx = ret_df.index
    n   = len(idx)
    out = pd.Series(np.nan, index=idx, name=f"AvgCorr_{window}w")

    if n < window:
        return out

    for i in range(window - 1, n):
        win = ret_df.iloc[i - window + 1 : i + 1]
        c   = win.corr()
        vals = c.values
        iu   = np.triu_indices_from(vals, k=1)
        if len(iu[0]) == 0:
            out.iloc[i] = np.nan
        else:
            out.iloc[i] = np.nanmean(vals[iu])
    return out

roll_corr_price = rolling_avg_corr(price_port, window=52)
roll_corr_total = rolling_avg_corr(total_port, window=52)

# =========================
# EQUITY CURVES & CHARTS
# =========================

def equity_curve(r, name):
    return (1 + r).cumprod().rename(name)

eq_price = pd.concat([
    equity_curve(rp_mtm_price,  "Port_MTM_price"),
    equity_curve(rp_cost_price, "Port_COST_price"),
    equity_curve(rb_jalsh_price,"ALSI_price"),
    equity_curve(rb_top40_price,"Top40_price"),
], axis=1)

eq_total = pd.concat([
    equity_curve(rp_mtm_total,  "Port_MTM_total"),
    equity_curve(rp_cost_total, "Port_COST_total"),
    equity_curve(rb_jalsh_total,"ALSI_total"),
    equity_curve(rb_top40_total,"Top40_total"),
], axis=1)

# Price-only equity curve
plt.figure(figsize=(10, 6))
eq_price.plot(ax=plt.gca())
plt.title(f"Equity Curves (Price Returns, {BACKFILL_START}{END_DATE})")
plt.ylabel("Growth of 1")
plt.xlabel("Date")
plt.grid(True)
plt.tight_layout()
plt.show()

# Total-return equity curve
plt.figure(figsize=(10, 6))
eq_total.plot(ax=plt.gca())
plt.title("Equity Curves (Total Returns with Synthetic Dividends)")
plt.ylabel("Growth of 1")
plt.xlabel("Date")
plt.grid(True)
plt.tight_layout()
plt.show()

# Rolling average correlation (price)
plt.figure(figsize=(10, 4))
roll_corr_price.plot()
plt.title("Rolling 52-week average pairwise correlation (price returns)")
plt.xlabel("Date")
plt.ylabel("Avg pairwise corr")
plt.grid(True)
plt.tight_layout()
plt.show()

# Rolling average correlation (total)
plt.figure(figsize=(10, 4))
roll_corr_total.plot()
plt.title("Rolling 52-week average pairwise correlation (total returns)")
plt.xlabel("Date")
plt.ylabel("Avg pairwise corr")
plt.grid(True)
plt.tight_layout()
plt.show()

# =========================
# OUTLIER REPORT
# =========================

print("\nNumber of flagged 100%+ weekly moves per ticker (treated as splits & median-filled):")
for t, flags in outlier_flags.items():
    print(f"{t}: {len(flags)}")

print("\nSanity check: weeks with |weekly return| > 100% AFTER cleaning (should be tiny / zero, mainly benchmarks if any):")
for t in price_panel.columns:
    big = price_panel[t].loc[price_panel[t].abs() > 1.0]
    print(f"{t}: {len(big)}")












```
