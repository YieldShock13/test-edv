```python
# ======================================================
# LEVEL-BASED RELATIVE VALUE (INDUSTRY STANDARD)
# Weekly data, 52-week rolling Z
# Pairwise RV + Multi-Asset Rich/Cheap
# ======================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ------------------------------
# Professional / neutral styling
# ------------------------------
plt.rcParams.update({
    "font.family": "serif",
    "font.serif": ["Times New Roman", "DejaVu Serif"],
    "axes.edgecolor": "#4A4A4A",
    "axes.labelcolor": "#2A2A2A",
    "xtick.color": "#2A2A2A",
    "ytick.color": "#2A2A2A",
    "text.color": "#2A2A2A",
    "figure.facecolor": "white",
    "axes.facecolor": "white"
})

WATERMARK = "Pearl Gray Equity and Research"

# ------------------------------
# Load prices
# ------------------------------
def load_prices(path, name):
    df = pd.read_csv(path)
    df["Date"] = pd.to_datetime(df["Date"])

    if "Price" in df.columns:
        col = "Price"
    elif "Close" in df.columns:
        col = "Close"
    else:
        raise ValueError(f"No price column in {name}")

    df = df[["Date", col]].rename(columns={col: name})
    df = df.sort_values("Date").set_index("Date")
    return df

spy = load_prices("SPY Historical Data (7).csv", "SPY")
gld = load_prices("GLD Historical Data (1).csv", "GLD")
xop = load_prices("XOP ETF Stock Price History.csv", "XOP")

# ------------------------------
# Align weekly prices
# ------------------------------
prices = pd.concat([spy, gld, xop], axis=1).dropna()

# ======================================================
# PAIRWISE RELATIVE VALUE (LEVEL-BASED)
# ======================================================
rv_xop_spy = np.log(prices["XOP"] / prices["SPY"])
rv_xop_gld = np.log(prices["XOP"] / prices["GLD"])

z_xop_spy = (rv_xop_spy - rv_xop_spy.rolling(52).mean()) / rv_xop_spy.rolling(52).std()
z_xop_gld = (rv_xop_gld - rv_xop_gld.rolling(52).mean()) / rv_xop_gld.rolling(52).std()

z_xop_spy = z_xop_spy.dropna()
z_xop_gld = z_xop_gld.dropna()

# ------------------------------
# Plot: XOP vs SPY
# ------------------------------
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(z_xop_spy.index, z_xop_spy, linewidth=2, color="#1F4E79")
ax.axhline(0, linestyle="--", linewidth=1, color="#8A8A8A")

ax.text(z_xop_spy.index[-1], z_xop_spy.iloc[-1],
        f"{z_xop_spy.iloc[-1]:.2f}", fontsize=9, color="#1F4E79",
        ha="left", va="center")

fig.text(0.5, 0.02, WATERMARK, ha="center", va="center", fontsize=9, color="black")

ax.set_title("Relative Value Z (Level): XOP vs SPY  52W Rolling", fontweight="bold")
ax.set_ylabel("Z-Score")
plt.tight_layout()
plt.show()

# ------------------------------
# Plot: XOP vs GLD
# ------------------------------
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(z_xop_gld.index, z_xop_gld, linewidth=2, color="#4F6F52")
ax.axhline(0, linestyle="--", linewidth=1, color="#8A8A8A")

ax.text(z_xop_gld.index[-1], z_xop_gld.iloc[-1],
        f"{z_xop_gld.iloc[-1]:.2f}", fontsize=9, color="#4F6F52",
        ha="left", va="center")

fig.text(0.5, 0.02, WATERMARK, ha="center", va="center", fontsize=9, color="black")

ax.set_title("Relative Value Z (Level): XOP vs GLD  52W Rolling", fontweight="bold")
ax.set_ylabel("Z-Score")
plt.tight_layout()
plt.show()

# ======================================================
# MULTI-ASSET RICH / CHEAP VS BASKET (LEVEL-BASED)
# ======================================================
geom_mean = np.exp(np.log(prices).mean(axis=1))
rv_multi = np.log(prices.div(geom_mean, axis=0))

z_multi = (rv_multi - rv_multi.rolling(52).mean()) / rv_multi.rolling(52).std()
z_multi = z_multi.dropna()

colors = {
    "SPY": "#1F4E79",
    "GLD": "#6E6E6E",
    "XOP": "#4F6F52"
}

fig, ax = plt.subplots(figsize=(11, 6))

for asset in z_multi.columns:
    s = z_multi[asset]
    ax.plot(s.index, s.values, linewidth=2, color=colors[asset], label=asset)
    ax.text(s.index[-1], s.values[-1],
            f"{asset}: {s.values[-1]:.2f}",
            fontsize=9, color=colors[asset],
            ha="left", va="center")

fig.text(0.5, 0.02, WATERMARK, ha="center", va="center", fontsize=9, color="black")

ax.axhline(0, linestyle="--", linewidth=1, color="#8A8A8A")
ax.set_title("Multi-Asset Rich / Cheap vs Basket (Level, 52W Rolling)", fontweight="bold")
ax.set_ylabel("Z-Score")
ax.legend(frameon=False)
plt.tight_layout()
plt.show()














```
