```python
import pandas as pd
import matplotlib.pyplot as plt

# --- Load MOVE ---
move = pd.read_csv("ICE BofAML MOVE Historical Data.csv")
move["Date"] = pd.to_datetime(move["Date"])
move["Price"] = move["Price"].astype(str).str.replace(",", "", regex=False).astype(float)
move = move[["Date", "Price"]].rename(columns={"Price": "MOVE"}).set_index("Date").sort_index()

# --- Load VIX ---
vix = pd.read_csv("VIXCLS (3).csv")
vix["observation_date"] = pd.to_datetime(vix["observation_date"])
vix = vix.rename(columns={"observation_date":"Date", "VIXCLS":"VIX"})
vix = vix[["Date","VIX"]].set_index("Date").astype(float).sort_index()

# --- Align & spread ---
df = move.join(vix, how="inner")
df["Spread"] = df["MOVE"] - df["VIX"]

# --- Plot helper ---
OCEAN_BG = "#0b1e2d"
LINE_COLOR = "#17becf"

def plot_spread(data, title):
    fig, ax = plt.subplots(figsize=(12,6))
    fig.patch.set_facecolor(OCEAN_BG)
    ax.set_facecolor(OCEAN_BG)

    ax.plot(data.index, data["Spread"], color=LINE_COLOR, linewidth=1.8)

    # Style
    ax.set_title(title, fontsize=16, fontweight="bold", color="white", pad=20)
    ax.set_ylabel("Spread (MOVE – VIX)", color="white")
    ax.set_xlabel("Date", color="white")
    ax.tick_params(colors="white")
    ax.spines["bottom"].set_color("white")
    ax.spines["left"].set_color("white")
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.grid(alpha=0.2, color="white")

    # Final data point
    last_date = data.index.max()
    last_val = float(data.loc[last_date, "Spread"])
    ax.scatter([last_date], [last_val], color="red", zorder=5)
    ax.annotate(f"{last_date.date()}: {last_val:.2f}",
                xy=(last_date, last_val),
                xytext=(8, 8),
                textcoords="offset points",
                fontsize=11, color="white", weight="bold",
                bbox=dict(facecolor=LINE_COLOR, alpha=0.4, edgecolor="none", boxstyle="round,pad=0.3"))

    # Watermark
    ax.text(0.5, 0.05, "Orca's Capital Strategies", transform=ax.transAxes,
            fontsize=18, color="white", alpha=0.15,
            ha="center", va="center", rotation=30)

    plt.tight_layout()
    plt.show()

# --- Chart 1: Full sample ---
plot_spread(df, "MOVE – VIX Spread (Full Sample)")

# --- Chart 2: Last 5 years ---
last5 = df[df.index >= (df.index.max() - pd.DateOffset(years=5))]
plot_spread(last5, "MOVE – VIX Spread (Last 5 Years)")
```
