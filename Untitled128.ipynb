```python
# ============================================================
# COLAB BACKTEST (MONTHLY, EXACT DATE MATCHING) + AUDIT + SMOOTHING TESTS
# Assets (5):
#  - DM ETF (SYGWDJ)  [price CSV from Investing.com]  + DIVIDENDS (manual, ex-div month)
#  - EM ETF (STXEMGJ) [price CSV from Investing.com]  (price only, no dividends supplied)
#  - Gold ETF (GLDJ)  [price CSV from Investing.com]  (price only, no dividends supplied)
#  - STX GOV ETF (STXGOVJ) [price CSV from Investing.com] + DIVIDENDS (manual, ex-div month)
#  - Peregrine Pure (manual monthly NET returns) [month-end already]
#
# Portfolios:
#  1) Equal weight across ALL 5
#  2) Long-only Max Sharpe (monthly) across ALL 5
#  3) Custom weights: 30% DM, 20% EM, 20% Peregrine, 15% STXGOV, 15% Gold
#
# Hard requirements implemented:
#  - Monthly returns only
#  - EXACT month-end date alignment using inner-join
#  - ETFs can be daily/weekly/monthly: we always compute monthly returns from month-end PRICE
#  - Peregrine returns are month-end already
#  - AUDIT/FAIL-FAST: file discovery, parse checks, month-end price sanity, stale/zero return checks,
#                     outlier moves check
#  - Metrics: CAGR, VaR95 (monthly), MaxDD, Annual Vol, Monthly Vol, TRR(CAGR/AnnVol),
#             Sharpe(mean/vol), correlations, equity curves + drawdowns
#  - Smoothing tests: ACF(1–6), Ljung–Box, HAC(Newey-West style) vol/sharpe vs plain
#  - Conditional stress: Peregrine during worst 10 DM months & worst 10 Custom months
#
# DIVIDENDS ADDED (only where you supplied them):
#  - STXGOVJ: YES (cash dividends -> monthly dividend return in ex-div month)
#  - SYGWDJ:  YES (cash dividends -> monthly dividend return in ex-div month)
#  - EM:      NO (not supplied)
#  - Gold:    NO (not supplied)
#  - Peregrine: already NET returns (unchanged)
# ============================================================

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display

# ---- Ensure statsmodels exists ----
try:
    from statsmodels.tsa.stattools import acf
    from statsmodels.stats.diagnostic import acorr_ljungbox
except Exception:
    !pip -q install statsmodels
    from statsmodels.tsa.stattools import acf
    from statsmodels.stats.diagnostic import acorr_ljungbox


# ============================================================
# 0) FILE DISCOVERY (NO MORE /mnt/data VS /content BS)
# ============================================================

FILENAMES = {
    "DM_FILE":     "SYGWDJ Historical Data.csv",
    "EM_FILE":     "STXEMGJ Historical Data.csv",
    "GOLD_FILE":   "GLDJ Historical Data.csv",
    "STXGOV_FILE": "STXGOVJ Historical Data (4).csv",
}

DM_NAME     = "DM (SYGWDJ)"
EM_NAME     = "EM (STXEMGJ)"
GOLD_NAME   = "Gold (GLDJ)"
STXGOV_NAME = "STXGOV (STXGOVJ)"
PERE_NAME   = "Peregrine (Net)"

OUTLIER_RETURN_THRESHOLD = 0.25    # flag monthly moves > 25%
STALE_PRICE_WARN_FRAC    = 0.20    # if >20% month-end prices repeat, warn
ZERO_RET_WARN_FRAC       = 0.20    # if >20% monthly returns are 0, warn

SEARCH_DIRS = [
    ".",                 # current
    "/content",           # colab default
    "/content/drive/MyDrive",
    "/mnt/data",          # some runtimes
    "/kaggle/input",      # if ever
]

def find_file(filename: str) -> str:
    for d in SEARCH_DIRS:
        p = os.path.join(d, filename)
        if os.path.exists(p):
            return p

    candidates = []
    for d in SEARCH_DIRS:
        if not os.path.isdir(d):
            continue
        try:
            for root, _, files in os.walk(d):
                for f in files:
                    if f == filename:
                        candidates.append(os.path.join(root, f))
                if root.count(os.sep) - d.count(os.sep) >= 3:
                    continue
        except Exception:
            pass
    if candidates:
        return candidates[0]

    raise FileNotFoundError(
        f"File not found: '{filename}'.\n"
        f"Searched: {SEARCH_DIRS}\n"
        f"Fix: ensure files are uploaded OR adjust SEARCH_DIRS."
    )

paths = {k: find_file(v) for k, v in FILENAMES.items()}
print("Resolved file paths:")
for k, p in paths.items():
    print(f"  {k}: {p}")


# ============================================================
# 1) PREPROCESSING: LOAD MONTH-END PRICES (AUDIT + FAIL FAST)
# ============================================================

def _parse_price_series(s: pd.Series) -> pd.Series:
    return (
        s.astype(str)
         .str.replace(",", "", regex=False)
         .str.strip()
         .replace({"": np.nan, "nan": np.nan, "None": np.nan})
         .astype(float)
    )

def _parse_date_series(s: pd.Series) -> pd.Series:
    dt = pd.to_datetime(s, errors="coerce", infer_datetime_format=True, dayfirst=False)
    if dt.isna().mean() > 0.30:
        dt2 = pd.to_datetime(s, errors="coerce", infer_datetime_format=True, dayfirst=True)
        if dt2.isna().mean() < dt.isna().mean():
            dt = dt2
    return dt

def month_end_index(d: pd.Series) -> pd.Series:
    return pd.to_datetime(d).dt.to_period("M").dt.to_timestamp("M")

def load_investing_price_csv_to_month_end_prices(path: str, label: str) -> pd.Series:
    if not os.path.exists(path):
        raise FileNotFoundError(f"{label}: file not found at: {path}")

    df = pd.read_csv(path)

    if "Date" not in df.columns:
        raise ValueError(f"{label}: missing 'Date' column. Found: {list(df.columns)}")
    if "Price" not in df.columns:
        raise ValueError(f"{label}: missing 'Price' column. Found: {list(df.columns)}")

    df = df.copy()
    df["Date"]  = _parse_date_series(df["Date"])
    df["Price"] = _parse_price_series(df["Price"])

    df = df.dropna(subset=["Date", "Price"]).sort_values("Date")
    if df.empty:
        raise ValueError(f"{label}: after parsing, no valid (Date, Price) rows remain.")

    df["MEnd"] = month_end_index(df["Date"])
    px_m = df.groupby("MEnd")["Price"].last().sort_index()
    px_m.name = label

    if not px_m.index.is_monotonic_increasing:
        raise ValueError(f"{label}: month-end index not monotonic increasing after grouping (unexpected).")

    stale_frac = float((px_m.diff() == 0).mean())
    if stale_frac > STALE_PRICE_WARN_FRAC:
        print(f"WARNING [{label}]: {stale_frac:.1%} of month-end prices are unchanged vs prior month.")

    return px_m

def prices_to_monthly_returns(px_m: pd.Series, label: str) -> pd.Series:
    r = px_m.pct_change()
    r.name = label
    zero_frac = float((r.dropna() == 0).mean()) if r.dropna().size else 0.0
    if zero_frac > ZERO_RET_WARN_FRAC:
        print(f"WARNING [{label}]: {zero_frac:.1%} of monthly returns are exactly 0.")
    return r

# Load month-end prices
dm_px     = load_investing_price_csv_to_month_end_prices(paths["DM_FILE"], DM_NAME)
em_px     = load_investing_price_csv_to_month_end_prices(paths["EM_FILE"], EM_NAME)
gold_px   = load_investing_price_csv_to_month_end_prices(paths["GOLD_FILE"], GOLD_NAME)
stxgov_px = load_investing_price_csv_to_month_end_prices(paths["STXGOV_FILE"], STXGOV_NAME)

# Price returns (unchanged)
dm_price     = prices_to_monthly_returns(dm_px, DM_NAME)
em_price     = prices_to_monthly_returns(em_px, EM_NAME)
gold_price   = prices_to_monthly_returns(gold_px, GOLD_NAME)
stxgov_price = prices_to_monthly_returns(stxgov_px, STXGOV_NAME)


# ============================================================
# 1B) DIVIDENDS: MANUAL CASH DIVS -> MONTHLY DIVIDEND RETURN
#     total_return_t = price_return_t + (div_cash_t / price_{t-1})
#     div_cash_t is assigned to the ex-div MONTH (month-end index)
# ============================================================

# --- Your supplied dividend events (ex-div date, cash dividend) ---
# STXGOVJ
STXGOV_DIV_EVENTS = [
    ("2025-10-22", 36.17),
    ("2025-07-16", 4.57),
    ("2025-04-15", 34.89),
    ("2025-01-15", 13.14),
    ("2024-10-16", 25.90),
    ("2024-07-17", 5.88),
    ("2024-04-24", 33.73),
    ("2024-01-17", 5.33),
    ("2023-10-18", 32.95),
    ("2023-07-26", 6.54),
    ("2023-04-19", 32.61),
    ("2023-01-18", 14.27),
    ("2022-10-19", 34.15),
    ("2022-08-03", 22.46),
    ("2022-01-19", 47.71),
    ("2021-07-14", 37.69),
    ("2021-01-13", 36.07),
    ("2020-07-22", 26.28),
]

# SYGWDJ (Developed markets)
DM_DIV_EVENTS = [
    ("2025-07-09", 32.02),
    ("2025-01-15", 23.21),
    ("2024-07-10", 32.78),
    ("2024-01-10", 25.37),
    ("2023-07-12", 32.62),
    ("2023-01-11", 22.01),
    ("2022-07-13", 26.70),
]

def _events_to_monthly_cash_series(events, label):
    df = pd.DataFrame(events, columns=["ex_date", "cash"])
    df["ex_date"] = pd.to_datetime(df["ex_date"])
    df["MEnd"] = df["ex_date"].dt.to_period("M").dt.to_timestamp("M")
    cash_m = df.groupby("MEnd")["cash"].sum().sort_index()
    cash_m.name = f"{label} cash div (sum)"
    return cash_m

def cash_div_to_monthly_div_return(cash_m: pd.Series, px_m: pd.Series, label: str) -> pd.Series:
    # dividend return uses prior month-end price as denominator: D_t / P_{t-1}
    denom = px_m.shift(1)
    div_ret = (cash_m / denom).reindex(px_m.index).fillna(0.0)
    div_ret.name = f"{label} dividend return"
    # first month is undefined in the same way price return is undefined
    div_ret.loc[div_ret.index.min()] = 0.0
    return div_ret

# Build monthly cash series
stxgov_div_cash_m = _events_to_monthly_cash_series(STXGOV_DIV_EVENTS, STXGOV_NAME)
dm_div_cash_m     = _events_to_monthly_cash_series(DM_DIV_EVENTS, DM_NAME)

# Convert to monthly dividend returns aligned to month-end price index
stxgov_div_ret = cash_div_to_monthly_div_return(stxgov_div_cash_m, stxgov_px, STXGOV_NAME)
dm_div_ret     = cash_div_to_monthly_div_return(dm_div_cash_m, dm_px, DM_NAME)

# Build TOTAL returns (price + dividend return)
stxgov = (stxgov_price.fillna(0.0) + stxgov_div_ret).rename(STXGOV_NAME)
dm     = (dm_price.fillna(0.0)     + dm_div_ret).rename(DM_NAME)

# Keep others unchanged (no dividends supplied)
em   = em_price.rename(EM_NAME)
gold = gold_price.rename(GOLD_NAME)

print("\nDIVIDENDS APPLIED (to total return series):")
print(f" - {STXGOV_NAME}: YES (cash, ex-div month) | events={len(STXGOV_DIV_EVENTS)}")
print(f" - {DM_NAME}:     YES (cash, ex-div month) | events={len(DM_DIV_EVENTS)}")
print(" - EM:            NO (none supplied)")
print(" - Gold:          NO (none supplied)")
print(" - Peregrine:     already NET returns (unchanged)")

# Audit tables: show months where dividend return > 0 and implied monthly dividend yield (D_t / P_{t-1})
def dividend_audit_table(cash_m, px_m, div_ret, title):
    df = pd.DataFrame({
        "CashDividend": cash_m.reindex(px_m.index).fillna(0.0),
        "PrevMonthEndPrice": px_m.shift(1),
        "DivReturn=D/Pprev": div_ret
    })
    df = df[df["CashDividend"] > 0].copy()
    df.index.name = "MonthEnd (ex-div month)"
    print(f"\n=== DIVIDEND AUDIT: {title} ===")
    if df.empty:
        print("No dividend months in the available price window.")
    else:
        display(df.style.format({
            "CashDividend":"{:.2f}",
            "PrevMonthEndPrice":"{:.2f}",
            "DivReturn=D/Pprev":"{:.3%}",
        }))

dividend_audit_table(stxgov_div_cash_m, stxgov_px, stxgov_div_ret, STXGOV_NAME)
dividend_audit_table(dm_div_cash_m, dm_px, dm_div_ret, DM_NAME)


# ============================================================
# 2) PEREGRINE MONTHLY NET RETURNS (MANUAL, MONTH-END)
# ============================================================

pere_data = {
    2020: [np.nan, -0.0108, 0.0031, 0.0221, 0.0220, 0.0227, 0.0092, 0.0158, -0.0090, 0.0305, -0.0120, 0.0014],
    2021: [0.0129, -0.0118, -0.0013, 0.0197, -0.0157, 0.0123, -0.0009, 0.0047, 0.0030, 0.0090, -0.0013, 0.0172],
    2022: [-0.0065, -0.0125, 0.0032, 0.0571, 0.0009, -0.0003, 0.0274, 0.0069, 0.0150, -0.0061, 0.0114, 0.0084],
    2023: [0.0213, 0.0093, 0.0039, 0.0103, -0.0060, 0.0136, 0.0103, 0.0169, 0.0013, 0.0018, 0.0265, 0.0085],
    2024: [0.0127, 0.0136, -0.0075, 0.0105, 0.0063, 0.0182, 0.0090, 0.0100, 0.0300, 0.0024, 0.0151, 0.0166],
    2025: [0.0103, -0.0054, 0.0130, 0.0107, 0.0121, 0.0057, 0.0086, 0.0009, 0.0189, 0.0049, 0.0008, np.nan],  # Dec missing
}

pere_idx, pere_ret = [], []
for y, months in pere_data.items():
    for m, val in enumerate(months, start=1):
        if pd.isna(val):
            continue
        dt = pd.Timestamp(year=y, month=m, day=1).to_period("M").to_timestamp("M")
        pere_idx.append(dt)
        pere_ret.append(val)

pere = pd.Series(pere_ret, index=pd.DatetimeIndex(pere_idx), name=PERE_NAME).sort_index()
if pere.index.duplicated().any():
    raise ValueError("Peregrine series has duplicate month-end dates (manual input error).")


# ============================================================
# 3) EXACT MONTHLY ALIGNMENT (INNER JOIN) + OUTLIER CHECK
# ============================================================

rets = pd.concat([dm, em, pere, stxgov, gold], axis=1)
rets = rets.dropna(how="any").sort_index()

print("\nEXACT common monthly window (all 5, inner-joined):")
print("Start:", rets.index.min(), "End:", rets.index.max(), "Rows(months):", len(rets))

flag = (rets.abs() > OUTLIER_RETURN_THRESHOLD)
if flag.any().any():
    print(f"\nFLAGGED MONTHLY MOVES (|ret| > {OUTLIER_RETURN_THRESHOLD:.0%}) IN COMMON WINDOW:")
    display(rets[flag.any(axis=1)])
else:
    print(f"\nNo monthly outliers > {OUTLIER_RETURN_THRESHOLD:.0%} in the common window.")

# extra audit for STXGOV specifically
stx = rets[STXGOV_NAME]
print("\nSTXGOV audit (common window):")
print("  Months:", len(stx))
print("  Zero-return fraction:", f"{(stx==0).mean():.1%}")
print("  Min/Max monthly return:", f"{stx.min():.2%} / {stx.max():.2%}")
print("  Month-end price repeats (full series):", f"{(stxgov_px.diff()==0).mean():.1%}")

# show first few rows (like your sanity check)
print("\nHEAD (common window returns):")
display(rets.head(8).style.format("{:.6f}"))


# ============================================================
# 4) METRICS + EQUITY CURVES
# ============================================================

def equity_curve(r: pd.Series) -> pd.Series:
    return (1.0 + r.fillna(0.0)).cumprod()

def max_drawdown_from_returns(r: pd.Series) -> float:
    eq = equity_curve(r)
    peak = eq.cummax()
    dd = (eq / peak) - 1.0
    return float(dd.min())

def metrics_from_monthly_returns(r: pd.Series) -> dict:
    r = r.dropna()
    n = int(len(r))
    if n < 3:
        return {
            "Months": n,
            "CAGR": np.nan,
            "VaR_95": np.nan,
            "Max_Drawdown": np.nan,
            "Monthly_Vol": np.nan,
            "Annual_Vol": np.nan,
            "TRR_CAGR_over_AnnualVol": np.nan,
            "Sharpe_Mean_over_Vol_Ann": np.nan,
        }

    years = n / 12.0
    growth = float((1.0 + r).prod())
    cagr = growth ** (1.0 / years) - 1.0

    mvol = float(r.std(ddof=1))
    avol = mvol * np.sqrt(12.0)

    var95 = float(np.quantile(r, 0.05))
    mdd = max_drawdown_from_returns(r)

    trr = (cagr / avol) if (np.isfinite(avol) and avol != 0) else np.nan

    mean_m = float(r.mean())
    mean_a = mean_m * 12.0
    sharpe = (mean_a / avol) if (np.isfinite(avol) and avol != 0) else np.nan

    return {
        "Months": n,
        "CAGR": cagr,
        "VaR_95": var95,
        "Max_Drawdown": mdd,
        "Monthly_Vol": mvol,
        "Annual_Vol": avol,
        "TRR_CAGR_over_AnnualVol": trr,
        "Sharpe_Mean_over_Vol_Ann": sharpe,
    }

def build_metrics_table(series_dict: dict) -> pd.DataFrame:
    rows = []
    for name, s in series_dict.items():
        m = metrics_from_monthly_returns(s)
        rows.append({"Name": name, **m})
    return pd.DataFrame(rows)


# ============================================================
# 5) SMOOTHING / AUTOCORR TESTS (PEREGRINE)
# ============================================================

def hac_annual_vol(series: pd.Series, lag: int = 3) -> float:
    x = (series.dropna() - series.dropna().mean()).values
    if len(x) < 5:
        return np.nan
    gamma0 = np.mean(x * x)
    var = gamma0
    for k in range(1, lag + 1):
        w = 1 - k / (lag + 1)
        gamma = np.mean(x[k:] * x[:-k])
        var += 2 * w * gamma
    return float(np.sqrt(var * 12.0))

print("\n====================")
print("PEREGRINE SMOOTHING TESTS")
print("====================")

pere_aligned = rets[PERE_NAME].dropna()

acf_vals = acf(pere_aligned, nlags=6, fft=False)
print("ACF lags 1–6:", acf_vals[1:])

lb = acorr_ljungbox(pere_aligned, lags=[1,2,3,4,5,6], return_df=True)
print("\nLjung–Box p-values:")
print(lb[["lb_stat", "lb_pvalue"]])

plain_ann_vol = float(pere_aligned.std(ddof=1) * np.sqrt(12.0))
hac_ann_vol   = float(hac_annual_vol(pere_aligned, lag=3))

mean_ann = float(pere_aligned.mean() * 12.0)
plain_sh = (mean_ann / plain_ann_vol) if plain_ann_vol > 0 else np.nan
hac_sh   = (mean_ann / hac_ann_vol) if (np.isfinite(hac_ann_vol) and hac_ann_vol > 0) else np.nan

print("\nVol + Sharpe comparison (rf=0):")
print(f"Plain annual vol : {plain_ann_vol:.2%}")
print(f"HAC annual vol   : {hac_ann_vol:.2%}")
print(f"Plain Sharpe     : {plain_sh:.2f}")
print(f"HAC Sharpe       : {hac_sh:.2f}")


# ============================================================
# 6) PORTFOLIOS (EQUAL, MAX SHARPE LONG-ONLY, CUSTOM)
# ============================================================

assets = list(rets.columns)
n = len(assets)

w_equal = np.repeat(1.0/n, n)

def max_sharpe_long_only(returns: pd.DataFrame) -> np.ndarray:
    mu = returns.mean().values
    cov = returns.cov().values

    try:
        from scipy.optimize import minimize

        def neg_sharpe(w):
            w = np.array(w)
            port_mu = float(w @ mu)
            port_vol = float(np.sqrt(w @ cov @ w))
            if port_vol <= 0:
                return 1e9
            return -(port_mu / port_vol)

        cons = ({"type": "eq", "fun": lambda w: np.sum(w) - 1.0},)
        bnds = [(0.0, 1.0) for _ in range(n)]
        x0 = np.repeat(1.0/n, n)

        res = minimize(neg_sharpe, x0=x0, bounds=bnds, constraints=cons, method="SLSQP")
        if not res.success:
            raise RuntimeError(res.message)

        w = np.clip(res.x, 0, 1)
        w = w / w.sum()
        return w

    except Exception:
        rng = np.random.default_rng(7)
        best_w = np.repeat(1.0/n, n)
        best_s = -1e9
        for _ in range(25000):
            w = rng.random(n)
            w = w / w.sum()
            port_mu = float(w @ mu)
            port_vol = float(np.sqrt(w @ cov @ w))
            if port_vol <= 0:
                continue
            s = port_mu / port_vol
            if s > best_s:
                best_s = s
                best_w = w
        return best_w

w_sharpe = max_sharpe_long_only(rets)

w_custom_map = {
    DM_NAME: 0.30,
    EM_NAME: 0.20,
    PERE_NAME: 0.20,
    STXGOV_NAME: 0.15,
    GOLD_NAME: 0.15,
}
w_custom = np.array([w_custom_map[c] for c in assets], dtype=float)
w_custom = w_custom / w_custom.sum()

port = pd.DataFrame(index=rets.index)
port["EqualWeight"] = rets.values @ w_equal
port["MaxSharpe_LongOnly"] = rets.values @ w_sharpe
port["Custom_30DM_20EM_20P_15GOV_15Gold"] = rets.values @ w_custom

print("\nWeights used:")
w_df = pd.DataFrame({"Asset": assets, "Equal": w_equal, "MaxSharpe": w_sharpe, "Custom": w_custom})
display(w_df.style.format({"Equal":"{:.2%}","MaxSharpe":"{:.2%}","Custom":"{:.2%}"}))


# ============================================================
# 7) CONDITIONAL STRESS CHECKS
# ============================================================

print("\n====================")
print("CONDITIONAL STRESS CHECKS")
print("====================")

worst_dm_idx = rets[DM_NAME].nsmallest(10).index
print("\nPeregrine in worst 10 DM months:")
display(rets.loc[worst_dm_idx, [DM_NAME, PERE_NAME]])

worst_custom_idx = port["Custom_30DM_20EM_20P_15GOV_15Gold"].nsmallest(10).index
print("\nWorst 10 Custom portfolio months (with component returns):")
display(pd.concat(
    [port.loc[worst_custom_idx, ["Custom_30DM_20EM_20P_15GOV_15Gold"]],
     rets.loc[worst_custom_idx, [DM_NAME, EM_NAME, GOLD_NAME, STXGOV_NAME, PERE_NAME]]],
    axis=1
))


# ============================================================
# 8) METRICS TABLES + CORRELATIONS
# ============================================================

asset_metrics = build_metrics_table({c: rets[c] for c in assets})
port_metrics  = build_metrics_table({c: port[c] for c in port.columns})

print("\n=== ASSET METRICS (monthly) ===")
display(asset_metrics.style.format({
    "CAGR":"{:.2%}","VaR_95":"{:.2%}","Max_Drawdown":"{:.2%}",
    "Monthly_Vol":"{:.2%}","Annual_Vol":"{:.2%}",
    "TRR_CAGR_over_AnnualVol":"{:.2f}",
    "Sharpe_Mean_over_Vol_Ann":"{:.2f}",
    "Months":"{:d}",
}))

print("\n=== PORTFOLIO METRICS (monthly) ===")
display(port_metrics.style.format({
    "CAGR":"{:.2%}","VaR_95":"{:.2%}","Max_Drawdown":"{:.2%}",
    "Monthly_Vol":"{:.2%}","Annual_Vol":"{:.2%}",
    "TRR_CAGR_over_AnnualVol":"{:.2f}",
    "Sharpe_Mean_over_Vol_Ann":"{:.2f}",
    "Months":"{:d}",
}))

print("\n=== CORRELATION (ASSETS, monthly returns, exact matched dates) ===")
display(rets.corr().style.format("{:.3f}"))

print("\n=== CORRELATION (PORTFOLIOS, monthly returns) ===")
display(port.corr().style.format("{:.3f}"))


# ============================================================
# 9) CHARTS: EQUITY CURVES + DRAWDOWNS (PORTFOLIOS + ASSETS)
# ============================================================

plt.figure(figsize=(12, 6))
for c in port.columns:
    plt.plot(port.index, equity_curve(port[c]), label=c)
plt.title("Equity Curves (Rebased to 1.0) — Portfolios (Monthly, Exact Date Match)")
plt.xlabel("Date")
plt.ylabel("Growth of 1.0")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

plt.figure(figsize=(12, 6))
for c in port.columns:
    eq = equity_curve(port[c])
    dd = (eq / eq.cummax()) - 1.0
    plt.plot(port.index, dd, label=c)
plt.title("Drawdowns — Portfolios (Monthly, Exact Date Match)")
plt.xlabel("Date")
plt.ylabel("Drawdown")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

plt.figure(figsize=(12, 6))
for c in assets:
    plt.plot(rets.index, equity_curve(rets[c]), label=c)
plt.title("Equity Curves (Rebased to 1.0) — Assets (Monthly, Exact Date Match)")
plt.xlabel("Date")
plt.ylabel("Growth of 1.0")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()





```
