```python
# ==========================================================
# Confluence Strategy (Monthly 1st): Yield z vs Futures z (hedged)
# LONG US when (z_yield < -THRESH) & (z_fut > +THRESH)
# LONG DE when (z_yield > +THRESH) & (z_fut < -THRESH)
# Else FLAT
#
# Runs 3 rounds: THRESH  {0.0, 0.5, 1.0}
# Outputs: z visuals, entry markers, equity, returns bars, drawdown,
# per-trade P&L, and metrics (CumGain, Win/Loss, VaR95, MaxDD, Ann.Vol, CAGR, TRR).
# Added: 20x leveraged track with per-roundtrip slippage on active months,
# levered CAGR, levered MaxDD, and 95% VaR (percent + money).
# Watermark: "Orca's Capital Strategies"
# ==========================================================

import os, glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import statsmodels.api as sm

# Try Colab auto-download
try:
    from google.colab import files as colab_files
    _COLAB = True
except Exception:
    _COLAB = False

# ---------------- Config ----------------
WATERMARK = "Orca's Capital Strategies"
DATE_FMT_M = DateFormatter("%Y-%m")
THRESH_LIST = [0.0, 0.5, 1.0]   # Round 1, 2, 3

# Filenames (monthly, stamped on the 1st)
FN_DE_YLD = "Germany 2-Year Bond Yield Historical Data (2).csv"
FN_US_YLD = "United States 2-Year Bond Yield Historical Data (1).csv"
FN_DE_FUT = "Euro SCHATZ Futures Historical Data.csv"
FN_US_FUT = "US 2 Year T-Note Futures Historical Data.csv"

# Leverage / cost params
LEVERAGE = 20.0                  # 20x
SLIPPAGE_BPS_PER_ROUNDTRIP = 2   # 2 bps per round trip (entry+exit) on notional
START_EQUITY = 100_000.0         # for money VaR and equity scaling if needed
PERIODS_PER_YEAR = 12            # monthly data

# ---------------- Helpers ----------------
def resolve_path(filename):
    for p in (f"/content/{filename}", f"/mnt/data/{filename}", filename):
        if os.path.isfile(p): return p
    for base in ("/content", "/mnt/data", "."):
        hits = glob.glob(os.path.join(base, "**", filename), recursive=True)
        if hits: return hits[0]
    raise FileNotFoundError(f"Could not find: {filename}")

def robust_date(s: pd.Series) -> pd.Series:
    s = s.astype(str)
    d1 = pd.to_datetime(s, dayfirst=True,  errors="coerce")
    d2 = pd.to_datetime(s, dayfirst=False, errors="coerce")
    return d1 if d1.max() >= d2.max() else d2

def parse_number(series: pd.Series) -> pd.Series:
    s = series.astype(str).str.strip().str.replace("%","",regex=False)
    smpl = s.sample(min(200, len(s)), random_state=0)
    comma_no_dot = ((smpl.str.contains(",")) & ~smpl.str.contains(r"\.")).mean()
    if comma_no_dot > 0.6:
        s = s.str.replace(".", "", regex=False)   # thousands dot
        s = s.str.replace(",", ".", regex=False)  # decimal comma  dot
    else:
        s = s.str.replace(",", "", regex=False)   # thousands comma
    return pd.to_numeric(s, errors="coerce")

def load_monthly_first(path: str, colname: str) -> pd.Series:
    df = pd.read_csv(path)
    df.columns = [c.strip() for c in df.columns]
    if "Date" not in df.columns or "Price" not in df.columns:
        raise ValueError(f"'Date' and 'Price' required in {path}. Found: {df.columns.tolist()}")
    df["Date"]  = robust_date(df["Date"])
    df["Price"] = parse_number(df["Price"])
    df = df.dropna(subset=["Date","Price"]).sort_values("Date").set_index("Date")
    # one obs per month (files are already 1st-of-month)
    df = df[~df.index.duplicated(keep="first")]
    return df["Price"].rename(colname)

def add_wm(fig): fig.text(0.01, 0.01, WATERMARK, fontsize=12, alpha=0.35, ha="left", va="bottom")
def annotate_last(ax, x, y, fmt="{:.2f}", dx=8, dy=0): ax.annotate(fmt.format(y[-1]), xy=(x[-1], y[-1]), xytext=(dx, dy), textcoords="offset points", va="center")
def finalize(fig): add_wm(fig); plt.tight_layout(); plt.show()

def equity_from_returns(inc: pd.Series, start=1.0) -> pd.Series:
    return pd.Series(start, index=inc.index).mul((1 + inc).cumprod())

def max_drawdown_from_equity(eq: pd.Series) -> float:
    roll_max = eq.cummax()
    dd = eq / roll_max - 1.0
    return dd.min(), dd

def cagr_from_returns(inc: pd.Series, periods_per_year=12) -> float:
    inc = inc.dropna()
    if inc.empty: return np.nan
    total = (1 + inc).prod()
    years = len(inc) / periods_per_year
    return total**(1/years) - 1

# ---------------- Load data ----------------
p_de_yld = resolve_path(FN_DE_YLD)
p_us_yld = resolve_path(FN_US_YLD)
p_de_fut = resolve_path(FN_DE_FUT)
p_us_fut = resolve_path(FN_US_FUT)

DE2Y = load_monthly_first(p_de_yld, "DE2Y_yld")
US2Y = load_monthly_first(p_us_yld, "US2Y_yld")
DE_F = load_monthly_first(p_de_fut, "DE_fut")
US_F = load_monthly_first(p_us_fut, "US_fut")

# Align indices
ix_yld = DE2Y.index.intersection(US2Y.index)
yld = pd.concat([DE2Y.reindex(ix_yld), US2Y.reindex(ix_yld)], axis=1).dropna()

ix_fut = DE_F.index.intersection(US_F.index)
fut = pd.concat([DE_F.reindex(ix_fut), US_F.reindex(ix_fut)], axis=1).dropna()

# ---------------- Yield spread (bps) + z ----------------
yld["Spread_bps"] = (yld["DE2Y_yld"] - yld["US2Y_yld"]) * 100.0
mu_y, sd_y = yld["Spread_bps"].mean(), yld["Spread_bps"].std(ddof=0)
yld["Z_yield"] = (yld["Spread_bps"] - mu_y) / sd_y if sd_y != 0 else np.nan

# ---------------- Futures (hedged) z ----------------
# 1) OLS hedge: DE_fut ~ const + beta * US_fut
Xf = sm.add_constant(fut["US_fut"])
yf = fut["DE_fut"]
ols_fut = sm.OLS(yf, Xf).fit()
beta_f = float(ols_fut.params["US_fut"])
alpha_f = float(ols_fut.params["const"])

# 2) Hedged spread and z
fut["Spread_hedged"] = fut["DE_fut"] - beta_f * fut["US_fut"]
mu_h, sd_h = fut["Spread_hedged"].mean(), fut["Spread_hedged"].std(ddof=0)
fut["Z_fut"] = (fut["Spread_hedged"] - mu_h) / sd_h if sd_h != 0 else np.nan

# ---------------- Forward returns (signal at t, hold to t+1) ----------------
DE_ret_fwd = DE_F.shift(-1) / DE_F - 1.0
US_ret_fwd = US_F.shift(-1) / US_F - 1.0
ret_df = pd.concat([DE_ret_fwd.rename("DE_ret_fwd"), US_ret_fwd.rename("US_ret_fwd")], axis=1)

# Common panel for strategy runs
panel_base = fut[["Z_fut","DE_fut","US_fut"]].join(yld[["Z_yield"]], how="inner").join(ret_df, how="inner").dropna()

# ---------------- Metrics (Unlevered & Levered) ----------------
def compute_metrics(unlev_ret: pd.Series, active_mask: pd.Series,
                    leverage=LEVERAGE, slippage_bps_rt=SLIPPAGE_BPS_PER_ROUNDTRIP,
                    start_equity=START_EQUITY, periods_per_year=PERIODS_PER_YEAR):
    """
    unlev_ret: strategy simple returns per period (decimal), 0 when flat.
    active_mask: True when a position is on for that month (one round-trip assumed).
    leverage: gross multiple on notional.
    slippage_bps_rt: round-trip slippage in basis points (on price).
    start_equity: for money VaR reporting.
    """
    unlev_ret = unlev_ret.fillna(0.0).astype(float)
    active = active_mask.reindex(unlev_ret.index).fillna(False)

    # Unlevered equity & stats (growth of $1)
    eq_unlev = equity_from_returns(unlev_ret, start=1.0)
    mdd_unlev, dd_unlev = max_drawdown_from_equity(eq_unlev)
    cagr_unlev = cagr_from_returns(unlev_ret, periods_per_year)
    ann_vol_unlev = unlev_ret.std(ddof=0) * np.sqrt(periods_per_year)
    var95_unlev = -np.nanpercentile(unlev_ret, 5)

    # Levered returns:
    # Apply leverage to returns; subtract slippage only in active months.
    # Slippage on equity = (slippage% of price) * LEVERAGE (since notional scales with leverage).
    slip_dec_rt = slippage_bps_rt / 10_000.0
    slip_equity_frac = slip_dec_rt * leverage
    slip_series = active.astype(float) * slip_equity_frac

    lev_gross = unlev_ret * leverage
    lev_net = lev_gross - slip_series

    eq_lev = equity_from_returns(lev_net, start=1.0)
    mdd_lev, dd_lev = max_drawdown_from_equity(eq_lev)
    cagr_lev = cagr_from_returns(lev_net, periods_per_year)
    ann_vol_lev = lev_net.std(ddof=0) * np.sqrt(periods_per_year)
    var95_lev_pct = -np.nanpercentile(lev_net, 5)              # one-period VaR (percent of equity)
    # Money VaR using rolling equity at start of period (more precise)
    eq_prev = eq_lev.shift(1).fillna(1.0) * start_equity
    pnl_money = lev_net * eq_prev
    var95_lev_money = -np.nanpercentile(pnl_money, 5)

    # Cum gains (growth of $1 - 1)
    cum_gain_unlev = eq_unlev.iloc[-1] - 1.0
    cum_gain_lev   = eq_lev.iloc[-1] - 1.0

    # Wins/losses based on unlevered monthly returns
    wins = (unlev_ret > 0).sum(); losses = (unlev_ret < 0).sum()
    wl = (wins / losses) if losses > 0 else np.nan

    # TRR = CAGR / Ann.Vol
    trr_unlev = (cagr_unlev / ann_vol_unlev) if ann_vol_unlev not in (None, 0) else np.nan
    trr_lev   = (cagr_lev   / ann_vol_lev)   if ann_vol_lev   not in (None, 0) else np.nan

    return {
        "unlev": {
            "ret": unlev_ret, "eq": eq_unlev, "dd": dd_unlev, "mdd": mdd_unlev,
            "cum_gain": cum_gain_unlev, "wl": wl, "var95_pct": var95_unlev,
            "ann_vol": ann_vol_unlev, "cagr": cagr_unlev, "trr": trr_unlev
        },
        "lev": {
            "ret_net": lev_net, "eq": eq_lev, "dd": dd_lev, "mdd": mdd_lev,
            "cum_gain": cum_gain_lev, "var95_pct": var95_lev_pct,
            "var95_money": var95_lev_money, "ann_vol": ann_vol_lev,
            "cagr": cagr_lev, "trr": trr_lev,
            "slip_series": slip_series
        }
    }

# ---------------- Strategy Runner ----------------
def run_confluence(thresh: float,
                   leverage=LEVERAGE,
                   slippage_bps_rt=SLIPPAGE_BPS_PER_ROUNDTRIP,
                   start_equity=START_EQUITY):
    """
    LONG US when (Z_yield <= -thresh) & (Z_fut >= +thresh)
    LONG DE when (Z_yield >= +thresh) & (Z_fut <= -thresh)
    Else FLAT
    One round-trip assumed in each active month (slippage applied then).
    """
    df = panel_base.copy()
    df["longUS"] = (df["Z_yield"] <= -thresh) & (df["Z_fut"] >= +thresh)
    df["longDE"] = (df["Z_yield"] >= +thresh) & (df["Z_fut"] <= -thresh)
    df["active"] = df["longUS"] | df["longDE"]

    strat = pd.Series(0.0, index=df.index, name="Strat_ret")
    strat[df["longUS"]] = df.loc[df["longUS"], "US_ret_fwd"]
    strat[df["longDE"]] = df.loc[df["longDE"], "DE_ret_fwd"]
    df["Strat_ret"] = strat

    # Metrics
    res = compute_metrics(df["Strat_ret"], df["active"],
                          leverage=leverage, slippage_bps_rt=slippage_bps_rt,
                          start_equity=start_equity, periods_per_year=PERIODS_PER_YEAR)

    un = res["unlev"]; lv = res["lev"]

    # --- Charts ---
    # (1) Z-yield and Z-futures with thresh lines
    figZ, axZ = plt.subplots(figsize=(11.2, 5.2))
    axZ.plot(df.index, df["Z_yield"], label="Yield z (DEUS, bps)")
    axZ.plot(df.index, df["Z_fut"],   label=f"Futures z (DE  US), ={beta_f:.3f}")
    axZ.axhline(0, linewidth=1)
    if thresh > 0:
        axZ.axhline(+thresh, linestyle="--", alpha=0.8, label=f"+{thresh}")
        axZ.axhline(-thresh, linestyle="--", alpha=0.8, label=f"{thresh}")
    axZ.set_title(f"Z-Scores  Confluence Strategy (threshold = {thresh})")
    axZ.set_ylabel("Z"); axZ.set_xlabel("Date"); axZ.grid(True, alpha=0.3)
    axZ.xaxis.set_major_formatter(DATE_FMT_M)
    annotate_last(axZ, df.index, df["Z_yield"].values, fmt="{:.2f}")
    annotate_last(axZ, df.index, df["Z_fut"].values,   fmt="{:.2f}", dy=-12)
    axZ.legend(); finalize(figZ)

    # (2) Futures prices with ENTRY markers
    figP, axP = plt.subplots(figsize=(12.0, 5.8))
    axP.plot(fut.index, fut["DE_fut"], label="DE futures")
    axP.plot(fut.index, fut["US_fut"], label="US futures")
    de_entries = df.index[df["longDE"]]
    us_entries = df.index[df["longUS"]]
    axP.scatter(de_entries, fut.loc[de_entries, "DE_fut"], marker="^", s=90, label="ENTRY: Long DE", zorder=5)
    axP.scatter(us_entries, fut.loc[us_entries, "US_fut"], marker="v", s=90, label="ENTRY: Long US", zorder=5)
    axP.set_title(f"Entries  Confluence Strategy (threshold = {thresh})")
    axP.set_ylabel("Price"); axP.set_xlabel("Date"); axP.grid(True, alpha=0.3)
    axP.xaxis.set_major_formatter(DATE_FMT_M)
    annotate_last(axP, fut.index, fut["DE_fut"].values, fmt="{:.2f}")
    annotate_last(axP, fut.index, fut["US_fut"].values, fmt="{:.2f}")
    axP.legend(); finalize(figP)

    # (3) Equity curve (unlevered vs levered)
    figE, axE = plt.subplots(figsize=(11.2, 5.2))
    axE.plot(df.index, un["eq"], label="Equity (Unlevered)")
    axE.plot(df.index, lv["eq"], label=f"Equity (Levered {LEVERAGE:.0f}x, net)")
    axE.set_title(f"Equity Curve  Confluence Strategy (threshold = {thresh})")
    axE.set_ylabel("Growth of $1"); axE.set_xlabel("Date"); axE.grid(True, alpha=0.3)
    axE.xaxis.set_major_formatter(DATE_FMT_M)
    annotate_last(axE, df.index, un["eq"].values, fmt="{:.3f}")
    annotate_last(axE, df.index, lv["eq"].values, fmt="{:.3f}", dy=-12)
    axE.legend(); finalize(figE)

    # (4) Monthly returns bars (unlevered)
    figR, axR = plt.subplots(figsize=(11.2, 4.2))
    colors = np.where(df["Strat_ret"] >= 0, "tab:blue", "tab:orange")
    axR.bar(df.index, df["Strat_ret"].values, width=20, align="center", color=colors)
    axR.axhline(0, linewidth=1)
    axR.set_title(f"Monthly Returns  Wins (blue) / Losses (orange)  threshold {thresh}")
    axR.set_ylabel("Return"); axR.set_xlabel("Date"); axR.grid(True, alpha=0.25)
    axR.xaxis.set_major_formatter(DATE_FMT_M)
    finalize(figR)

    # (5) Drawdown (unlevered vs levered)
    figD, axD = plt.subplots(figsize=(11.2, 3.8))
    axD.plot(df.index, un["dd"], label="Drawdown (Unlevered)")
    axD.plot(df.index, lv["dd"], label=f"Drawdown (Levered {LEVERAGE:.0f}x)")
    axD.set_title(f"Drawdown  Confluence Strategy (threshold = {thresh})")
    axD.set_ylabel("Drawdown"); axD.set_xlabel("Date"); axD.grid(True, alpha=0.3)
    axD.xaxis.set_major_formatter(DATE_FMT_M)
    axD.legend(); finalize(figD)

    # Per-trade P&L (unlevered returns shown per your original)
    trades = df.loc[df["active"], ["Z_yield","Z_fut","DE_ret_fwd","US_ret_fwd","Strat_ret"]].copy()
    trades["Leg"] = np.where(df.loc[trades.index, "longUS"], "US", "DE")
    trades.index.name = "Date"

    # --- Print metrics ---
    print(f"\n=== Confluence Strategy  threshold {thresh} ===")
    print(f" (hedge) from OLS: {beta_f:.6f}  |  : {alpha_f:.6f}")
    print(f"Months: {len(df)}  |  Active months: {df['active'].sum()}")
    print("--- Unlevered ---")
    print(f"Cumulative Gain: {un['cum_gain']:.2%}")
    print(f"Max Drawdown: {un['mdd']:.2%}")
    print(f"VaR 95% (1-month): {un['var95_pct']:.2%}")
    print(f"Annualized Volatility: {un['ann_vol']:.2%}")
    print(f"CAGR: {un['cagr']:.2%}  |  TRR: {un['trr']:.3f}")
    print("--- Levered ({:.0f}x, slippage {} bps RT on active months) ---".format(LEVERAGE, SLIPPAGE_BPS_PER_ROUNDTRIP))
    print(f"Cumulative Gain: {lv['cum_gain']:.2%}")
    print(f"Max Drawdown: {lv['mdd']:.2%}")
    print(f"VaR 95% (1-month): {lv['var95_pct']:.2%}")
    print(f"VaR 95% (money, one-period): {lv['var95_money']:,.2f}")
    print(f"Annualized Volatility: {lv['ann_vol']:.2%}")
    print(f"CAGR: {lv['cagr']:.2%}  |  TRR: {lv['trr']:.3f}")
    print("\n--- Per-Trade P&L (unlevered monthly returns) ---")
    print(trades[["Leg","Z_yield","Z_fut","DE_ret_fwd","US_ret_fwd","Strat_ret"]]
          .to_string(float_format=lambda x: f"{x:,.4f}"))

    # --- Save outputs ---
    out = pd.concat(
        [
            yld[["Spread_bps","Z_yield"]],
            fut[["Spread_hedged","Z_fut"]],
            df[["DE_ret_fwd","US_ret_fwd","Strat_ret","active"]],
            un["eq"].rename("Equity_Unlev"),
            un["dd"].rename("Drawdown_Unlev"),
            lv["ret_net"].rename("Ret_Levered_Net"),
            lv["eq"].rename("Equity_Lev"),
            lv["dd"].rename("Drawdown_Lev"),
            lv["slip_series"].rename("Slippage_Frac_on_Equity")
        ],
        axis=1
    )
    out_path = f"/content/confluence_strategy_thresh_{str(thresh).replace('.','p')}.csv"
    out.to_csv(out_path)
    print(f"\nSaved CSV -> {out_path}")
    if _COLAB:
        try:
            colab_files.download(out_path)
        except Exception:
            pass

# ---------------- RUN all rounds ----------------
print("Confluence strategy confirmation:")
print(" LONG US when (z_yield <= -THRESH) & (z_fut >= +THRESH)")
print(" LONG DE when (z_yield >= +THRESH) & (z_fut <= -THRESH)")
print(" Else FLAT, rebalance monthly on the 1st (hold to next month).")

for thr in THRESH_LIST:
    run_confluence(thr)

```
