```python
# === IG Effective Yield  10Y Treasury (bps): Nearest-Available Last Obs per Quarter ===
# Colab-ready. End date fixed to 2025-08-27. Dark theme, watermark "Orcas Spread".

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ---------- CONFIG ----------
IG_EY_CSV_PATH = "BAMLC0A0CMEY.csv"    # IG Effective Yield (ICE/BofA)
TN_CSV_PATH    = "DGS10 (8).csv"       # 10Y Treasury yield (FRED)
TARGET_END_DATE = "2025-08-27"

SERIES_NAME_IGEY = "IG Effective Yield"
SERIES_NAME_10   = "US 10Y"
SPREAD_NAME      = "IG EY  10Y (bps)"
Z_NAME           = "Z-Score (IG EY  10Y)"
WATERMARK_TEXT   = "Orcas Spread"

DPI = 160
FONT_FAMILY = "DejaVu Sans"
plt.rcParams.update({
    "font.family": FONT_FAMILY,
    "figure.facecolor": "black",
    "axes.facecolor": "black",
    "savefig.facecolor": "black",
    "text.color": "white",
    "axes.labelcolor": "white",
    "xtick.color": "white",
    "ytick.color": "white",
    "axes.grid": True,
    "grid.color": "#555555",
    "grid.alpha": 0.25
})
# ---------------------------

def read_series_to_bps(path):
    """Read CSV, detect date/value columns, convert to bps if needed."""
    df = pd.read_csv(path)
    # detect date col
    date_col = None
    for c in df.columns:
        try:
            pd.to_datetime(df[c])
            date_col = c
            break
        except Exception:
            continue
    if date_col is None:
        raise ValueError(f"No parseable date column in {path}")

    # detect numeric col
    numeric_cols = [c for c in df.columns if pd.api.types.is_numeric_dtype(df[c]) and c != date_col]
    if not numeric_cols:
        for c in df.columns:
            if c != date_col:
                df[c] = pd.to_numeric(df[c].astype(str).str.replace(",",""), errors="coerce")
        numeric_cols = [c for c in df.columns if pd.api.types.is_numeric_dtype(df[c]) and c != date_col]
        if not numeric_cols:
            raise ValueError(f"No numeric value column in {path}")
    val_col = numeric_cols[-1]

    s = pd.Series(pd.to_numeric(df[val_col], errors="coerce").values,
                  index=pd.to_datetime(df[date_col]),
                  name=val_col).dropna().sort_index()

    # convert to bps
    p95 = float(s.quantile(0.95))
    if p95 <= 1.0:        # fraction -> bps
        s = s * 10000.0
    elif p95 <= 20.0:     # percent -> bps
        s = s * 100.0
    # else already bps
    return s

def nearest_last_obs(series: pd.Series) -> pd.Series:
    """Last actual obs per quarter, timestamped at actual date."""
    last_idx = series.groupby(series.index.to_period("Q")).apply(lambda x: x.index.max())
    vals = series.loc[last_idx.values].values
    return pd.Series(vals, index=last_idx.values, name=series.name).sort_index()

def zscore(series: pd.Series):
    mu = series.mean()
    sd = series.std(ddof=1)
    return (series - mu) / sd

def add_watermark(fig, text=WATERMARK_TEXT, alpha=0.25):
    fig.text(0.995, 0.02, text, fontsize=10, color="white",
             ha="right", va="bottom", alpha=alpha)

# --- Load data ---
igey_daily = read_series_to_bps(IG_EY_CSV_PATH).rename(SERIES_NAME_IGEY)
tn_daily   = read_series_to_bps(TN_CSV_PATH).rename(SERIES_NAME_10)
daily = pd.concat([igey_daily, tn_daily], axis=1).dropna().sort_index()

# restrict to <= target end date
daily = daily.loc[:TARGET_END_DATE]

# --- Nearest-available last obs per quarter ---
igey_qn = nearest_last_obs(daily[SERIES_NAME_IGEY])
tn_qn   = nearest_last_obs(daily[SERIES_NAME_10])
qe_df   = pd.concat([igey_qn, tn_qn], axis=1).dropna().sort_index()

spread_qn = (qe_df[SERIES_NAME_IGEY] - qe_df[SERIES_NAME_10]).rename(SPREAD_NAME)
z_qn = zscore(spread_qn).rename(Z_NAME)

# --- Plot Spread ---
fig1, ax1 = plt.subplots(figsize=(11,4), dpi=DPI)
ax1.plot(spread_qn.index, spread_qn.values, linewidth=2, marker="o", markersize=3)
ax1.set_title("IG Effective Yield  10Y Treasury (Nearest available last obs per quarter)", pad=12)
ax1.set_ylabel("Basis Points")
ax1.set_xlabel("Date")
ax1.axhline(0, linestyle="--", alpha=0.5)
ax1.annotate(f"{spread_qn.iloc[-1]:.0f} bps ({spread_qn.index[-1].date()})",
             xy=(spread_qn.index[-1], spread_qn.iloc[-1]),
             xytext=(10,0), textcoords="offset points", va="center")
add_watermark(fig1)
plt.tight_layout()
plt.show()

# --- Plot Z-Score ---
fig2, ax2 = plt.subplots(figsize=(11,4), dpi=DPI)
ax2.plot(z_qn.index, z_qn.values, linewidth=2, marker="o", markersize=3)
ax2.set_title("IG Effective Yield  10Y Treasury Spread  Z-Score", pad=12)
ax2.set_ylabel("Z")
ax2.set_xlabel("Date")
ax2.axhline(0, linestyle="--", alpha=0.5)
for thr, ls in [(1,"-."),(2,":")]:
    ax2.axhline(thr, linestyle=ls, alpha=0.6)
    ax2.axhline(-thr, linestyle=ls, alpha=0.6)
ax2.fill_between(z_qn.index, 1, z_qn, where=z_qn>=1, alpha=0.12)
ax2.fill_between(z_qn.index, 2, z_qn, where=z_qn>=2, alpha=0.18)
ax2.fill_between(z_qn.index, -1, z_qn, where=z_qn<=-1, alpha=0.12)
ax2.fill_between(z_qn.index, -2, z_qn, where=z_qn<=-2, alpha=0.18)
ax2.annotate(f"{z_qn.iloc[-1]:.2f}  ({z_qn.index[-1].date()})",
             xy=(z_qn.index[-1], z_qn.iloc[-1]),
             xytext=(10,0), textcoords="offset points", va="center")
add_watermark(fig2)
plt.tight_layout()
plt.show()

# --- Save outputs ---
qe_df.assign(**{SPREAD_NAME: spread_qn}).to_csv("igey_10y_nearest_last_obs_levels_and_spread.csv")
z_qn.to_frame().to_csv("igey_10y_nearest_last_obs_zscore.csv")


```
