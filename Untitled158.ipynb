```python
# ============================================================
# FULL WRITE-UP RISK ENGINE + BENCHMARK + EVT DIAGNOSTICS
# (Styled + Watermark Version)
# ============================================================

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import genpareto, probplot

# ------------------------------------------------------------
# PROFESSIONAL STYLING (NO DEFAULT BLUE/ORANGE)
# ------------------------------------------------------------

plt.style.use("seaborn-v0_8-white")

plt.rcParams.update({
    "figure.figsize": (12,6),
    "axes.titlesize": 14,
    "axes.labelsize": 12,
    "font.size": 11,
    "axes.edgecolor": "black",
    "axes.linewidth": 1.2
})

PRIMARY = "#1A1A1A"      # deep charcoal
SECONDARY = "#4C4C4C"    # muted grey
ACCENT = "#7A7A7A"       # light grey

def add_watermark(ax):
    ax.text(
        0.99, 0.02,
        "Orca's Capital Strategies",
        transform=ax.transAxes,
        fontsize=9,
        color="grey",
        alpha=0.35,
        ha="right",
        va="bottom"
    )

# ------------------------------------------------------------
# 1) LOAD DATA
# ------------------------------------------------------------

FILES = {
    "BRK": ["BRKb Historical Data (1).csv", "/mnt/data/BRKb Historical Data (1).csv"],
    "GLD": ["GLD Historical Data (2).csv",  "/mnt/data/GLD Historical Data (2).csv"],
    "USO": ["USO Historical Data (1).csv",  "/mnt/data/USO Historical Data (1).csv"],
}

def resolve_path(paths):
    for p in paths:
        if os.path.exists(p):
            return p
    raise FileNotFoundError

def load_price(path):
    df = pd.read_csv(path)
    df.columns = [c.strip() for c in df.columns]
    df["Date"] = pd.to_datetime(df["Date"])
    df["Price"] = df["Price"].astype(str).str.replace(",", "", regex=False).astype(float)
    df = df.sort_values("Date").set_index("Date")
    return df["Price"]

prices = {}
for k, v in FILES.items():
    prices[k] = load_price(resolve_path(v))

prices = pd.concat(prices, axis=1).dropna()
prices = prices.loc[prices.index >= prices.index.max() - pd.DateOffset(years=10)]

rets = prices.pct_change().dropna()

# ------------------------------------------------------------
# 2) TRUE MONTHLY DRIFT + REBALANCE
# ------------------------------------------------------------

target_w = pd.Series({"BRK":0.6, "GLD":0.2, "USO":0.2})

weights = []
port_returns = []
turnover = []
current_w = target_w.copy()

monthly_dates = rets.resample("ME").last().index

for date in rets.index:
    r = rets.loc[date]
    port_r = np.dot(current_w, r)
    port_returns.append(port_r)

    new_w = current_w * (1 + r)
    new_w = new_w / new_w.sum()

    if date in monthly_dates:
        trade = target_w - new_w
        turnover.append(np.abs(trade).sum())
        current_w = target_w.copy()
    else:
        turnover.append(0.0)
        current_w = new_w

    weights.append(current_w.values)

port_ret = pd.Series(port_returns, index=rets.index)
turnover = pd.Series(turnover, index=rets.index)

# ------------------------------------------------------------
# 3) RISK METRICS FUNCTION
# ------------------------------------------------------------

def risk_metrics(r):
    ann_vol = r.std() * np.sqrt(252)
    var95 = np.quantile(r, 0.05)
    var99 = np.quantile(r, 0.01)
    es95 = r[r <= var95].mean()
    es99 = r[r <= var99].mean()
    cumulative = (1 + r).cumprod()
    max_dd = (cumulative / cumulative.cummax() - 1).min()
    return pd.Series({
        "Ann Vol": ann_vol,
        "Hist VaR 95": var95,
        "Hist ES 95": es95,
        "Hist VaR 99": var99,
        "Hist ES 99": es99,
        "Max Drawdown": max_dd
    })

# ------------------------------------------------------------
# 4) BASELINE + BENCHMARK
# ------------------------------------------------------------

baseline_full = risk_metrics(port_ret)
brk_ret = rets["BRK"]
baseline_brk = risk_metrics(brk_ret)

# ------------------------------------------------------------
# 5) VOL REGIME
# ------------------------------------------------------------

rolling_vol = port_ret.rolling(63).std() * np.sqrt(252)
vol_cutoff = rolling_vol.quantile(0.70)

high_vol_ret = port_ret[rolling_vol >= vol_cutoff]
low_vol_ret  = port_ret[rolling_vol < vol_cutoff]

baseline_high = risk_metrics(high_vol_ret)
baseline_low  = risk_metrics(low_vol_ret)

# ------------------------------------------------------------
# 6) EVT
# ------------------------------------------------------------

def fit_evt(series):
    loss = -series.dropna()
    threshold = np.quantile(loss, 0.95)
    excess = loss[loss > threshold] - threshold

    if len(excess) < 30:
        return None

    c, loc, scale = genpareto.fit(excess, floc=0)
    p_u = len(excess) / len(loss)
    alpha = 0.05

    x = threshold + (scale/c)*((alpha/p_u)**(-c)-1) if abs(c)>1e-6 else threshold + scale*np.log(p_u/alpha)
    var_evt = -x
    es_evt = None

    if c < 1:
        es_loss = x + (scale + c*(x-threshold))/(1-c)
        es_evt = -es_loss

    return {
        "VaR": var_evt,
        "ES": es_evt,
        "Shape": c,
        "Threshold": threshold,
        "Excess": excess,
        "Scale": scale
    }

evt_full = fit_evt(port_ret)
evt_brk = fit_evt(brk_ret)

# ------------------------------------------------------------
# 7) LIQUIDITY OVERLAY
# ------------------------------------------------------------

liq_levels = [0.0005, 0.0025, 0.01]
liq_results = {}

for bps in liq_levels:
    liquidity_cost = turnover * bps
    adj_ret = port_ret - liquidity_cost
    liq_results[f"{int(bps*10000)} bps"] = risk_metrics(adj_ret)

liq_results = pd.DataFrame(liq_results)

# ------------------------------------------------------------
# 8) STRESS WINDOWS
# ------------------------------------------------------------

stress = pd.Series({
    "Worst 1d": port_ret.min(),
    "Worst 21d": port_ret.rolling(21).sum().min(),
    "Worst 63d": port_ret.rolling(63).sum().min()
})

# ------------------------------------------------------------
# 9) CHARTS (NEW COLOUR SCHEME + WATERMARK)
# ------------------------------------------------------------

# Equity Curve
eq_port = (1 + port_ret).cumprod()
eq_brk  = (1 + brk_ret).cumprod()

fig, ax = plt.subplots()
ax.plot(eq_port, color=PRIMARY, linewidth=2, label="Portfolio 60/20/20")
ax.plot(eq_brk, color=SECONDARY, linestyle="--", label="100% BRK")
ax.set_title("Equity Curve Comparison")
ax.legend()
add_watermark(ax)
plt.show()

# Drawdown
dd_port = eq_port / eq_port.cummax() - 1
dd_brk  = eq_brk / eq_brk.cummax() - 1

fig, ax = plt.subplots()
ax.plot(dd_port, color=PRIMARY, linewidth=2, label="Portfolio")
ax.plot(dd_brk, color=SECONDARY, linestyle="--", label="BRK")
ax.set_title("Drawdown Comparison")
ax.legend()
add_watermark(ax)
plt.show()

# Rolling Vol
fig, ax = plt.subplots()
ax.plot(rolling_vol, color=PRIMARY, label="63d Rolling Vol")
ax.axhline(vol_cutoff, linestyle="--", color=SECONDARY, label="High-Vol Threshold")
ax.set_title("Volatility Regime Split")
ax.legend()
add_watermark(ax)
plt.show()

# EVT Histogram
if evt_full:
    loss = -port_ret.dropna()
    threshold = evt_full["Threshold"]
    excess = evt_full["Excess"]
    c = evt_full["Shape"]
    scale = evt_full["Scale"]

    fig, ax = plt.subplots()
    sns.histplot(loss, bins=50, stat="density", color=ACCENT, ax=ax)
    x = np.linspace(threshold, loss.max(), 200)
    y = genpareto.pdf(x - threshold, c, scale=scale)
    ax.plot(x, y, color=PRIMARY, linewidth=2)
    ax.axvline(threshold, linestyle="--", color=SECONDARY)
    ax.set_title("EVT Tail Fit (Full Sample)")
    add_watermark(ax)
    plt.show()

# Mean Excess Plot
thresholds = np.linspace(loss.quantile(0.80), loss.quantile(0.99), 30)
mean_excess = [(loss[loss > u] - u).mean() for u in thresholds]

fig, ax = plt.subplots()
ax.plot(thresholds, mean_excess, color=PRIMARY, linewidth=2)
ax.set_title("Mean Excess Function")
add_watermark(ax)
plt.show()

# QQ Plot
fig, ax = plt.subplots()
probplot(excess, dist="genpareto",
         sparams=(evt_full["Shape"],0,evt_full["Scale"]), plot=ax)
ax.set_title("EVT QQ Plot")
add_watermark(ax)
plt.show()

# ------------------------------------------------------------
# OUTPUT TABLES
# ------------------------------------------------------------

print("=== PORTFOLIO BASELINE ===")
print(baseline_full)

print("\n=== BRK BASELINE ===")
print(baseline_brk)

print("\n=== EVT PORTFOLIO ===")
print(evt_full)

print("\n=== EVT BRK ===")
print(evt_brk)

print("\n=== LIQUIDITY STRESS ===")
print(liq_results)

print("\n=== STRESS WINDOWS ===")
print(stress)


```
