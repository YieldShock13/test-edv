```python
# For Google Colab or fresh environments
!pip install requests matplotlib pandas

import pandas as pd
import requests
import matplotlib.pyplot as plt
import numpy as np

# --- FRED API details ---
api_key = "83cf5af9953ef78768ee3b3186db216b"
series_id = "SP500"
url = "https://api.stlouisfed.org/fred/series/observations"
params = {
    "series_id": series_id,
    "api_key": api_key,
    "file_type": "json",
    "observation_start": "2005-01-01"
}

# --- Download from FRED ---
response = requests.get(url, params=params)
data = response.json()["observations"]
df = pd.DataFrame(data)
df["date"] = pd.to_datetime(df["date"])
df["value"] = pd.to_numeric(df["value"], errors="coerce")
df = df.dropna()

# --- Plot 1: S&P 500 + 30-day SMA ---
df["30d_SMA"] = df["value"].rolling(window=30).mean()

fig, ax = plt.subplots(figsize=(14, 6))
fig.patch.set_facecolor('black')
ax.plot(df["date"], df["value"], label="S&P 500", alpha=0.5, color='gray')
ax.plot(df["date"], df["30d_SMA"], label="30-day SMA", color="white", linewidth=2)

# Data labels
latest_date = df["date"].iloc[-1]
ax.text(latest_date, df["value"].iloc[-1], f"{df['value'].iloc[-1]:.2f}", color="gray", fontsize=10, ha="left", va="center")
ax.text(latest_date, df["30d_SMA"].iloc[-1], f"{df['30d_SMA'].iloc[-1]:.2f}", color="white", fontsize=10, ha="left", va="center")

# Watermark outside plot (top-left corner)
fig.text(0.01, 0.99, "Pearl Gray Equity and Research", ha='left', va='top',
         fontsize=13, color='white', fontweight='bold')

ax.set_title("S&P 500 Index & 30-day SMA (2005–Present)", fontsize=14, weight='bold')
ax.set_xlabel("Date")
ax.set_ylabel("Index Level")
ax.grid(True, linestyle='--', alpha=0.2)
ax.legend()
plt.style.use('dark_background')
plt.tight_layout()
plt.show()

# --- Plot 2: Z-score vs 12mo mean ---
df["12mo_mean"] = df["value"].rolling(window=252).mean()
df["12mo_std"] = df["value"].rolling(window=252).std()
df["Z_score"] = (df["value"] - df["12mo_mean"]) / df["12mo_std"]

fig, ax = plt.subplots(figsize=(14, 6))
fig.patch.set_facecolor('black')
ax.plot(df["date"], df["Z_score"], label="Z-Score", color="cyan")
ax.axhline(2, color="red", linestyle="--", linewidth=1, label="+2σ")
ax.axhline(-2, color="red", linestyle="--", linewidth=1, label="-2σ")
ax.axhline(0, color="white", linestyle="--", linewidth=0.5)

# Z-score label
ax.text(latest_date, df["Z_score"].iloc[-1], f"{df['Z_score'].iloc[-1]:.2f}", color="cyan", fontsize=10, ha="left", va="center")

# Watermark outside plot (top-left corner)
fig.text(0.01, 0.99, "Pearl Gray Equity and Research", ha='left', va='top',
         fontsize=13, color='white', fontweight='bold')

ax.set_title("S&P 500 Z-Score (vs 12-mo Mean)", fontsize=14, weight='bold')
ax.set_xlabel("Date")
ax.set_ylabel("Z-score")
ax.grid(True, linestyle='--', alpha=0.2)
ax.legend()
plt.style.use('dark_background')
plt.tight_layout()
plt.show()
























```
