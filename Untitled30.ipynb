```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
from datetime import datetime

# === API key ===
api_key = "83cf5af9953ef78768ee3b3186db216b"

# === Fetch FRED data ===
def fetch_fred_series(series_id, start="2005-01-01", end=None):
    if end is None:
        end = datetime.today().strftime("%Y-%m-%d")
    url = "https://api.stlouisfed.org/fred/series/observations"
    params = {
        "series_id": series_id,
        "api_key": api_key,
        "file_type": "json",
        "observation_start": start,
        "observation_end": end,
        "frequency": "d"
    }
    r = requests.get(url, params=params)
    data = r.json()["observations"]
    df = pd.DataFrame(data)
    df["date"] = pd.to_datetime(df["date"])
    df.set_index("date", inplace=True)
    df = pd.to_numeric(df["value"], errors="coerce")
    return df.dropna()

# === Fetch VIX ===
vix = fetch_fred_series("VIXCLS")
vix_sma = vix.rolling(30).mean().rename("30-day SMA")

# === Full Chart ===
plt.figure(figsize=(12, 6), facecolor='black')
ax1 = plt.gca()
ax1.set_facecolor('black')

vix.plot(ax=ax1, color='gray', linewidth=1.5, label="VIX")
vix_sma.plot(ax=ax1, color='white', linewidth=2, label="30-day SMA")

# Label final VIX value, shifted right
last_date = vix.index[-1]
last_val = vix.iloc[-1]
ax1.text(last_date + pd.Timedelta(days=20), last_val,
         f"{last_val:.2f}", color='white', fontsize=9, fontweight='bold',
         ha='left', va='center')

# Watermark
ax1.text(0.01, 1.08, 'Pearl Gray Equity and Research', transform=ax1.transAxes,
         fontsize=10, color='white', alpha=0.8, fontweight='bold', ha='left', va='top')

# Style
ax1.set_title("VIX Index & 30-day SMA (2005Present)", fontsize=14, fontweight='bold', color='white')
ax1.set_xlabel("Date", fontsize=12, color='white', fontweight='bold')
ax1.set_ylabel("VIX Level", fontsize=12, color='white', fontweight='bold')
ax1.legend(facecolor='black', edgecolor='white', labelcolor='white', fontsize=10)
ax1.grid(True, linestyle='--', color='white', alpha=0.2)
ax1.tick_params(colors='white')
for spine in ax1.spines.values():
    spine.set_color('white')

plt.tight_layout()
plt.show()

# === YTD Chart ===
vix_ytd = vix[vix.index >= pd.Timestamp(f"{datetime.today().year}-01-01")]
vix_sma_ytd = vix_ytd.rolling(30).mean().rename("30-day SMA (YTD)")

plt.figure(figsize=(12, 6), facecolor='black')
ax2 = plt.gca()
ax2.set_facecolor('black')

vix_ytd.plot(ax=ax2, color='gray', linewidth=1.5, label="VIX (YTD)")
vix_sma_ytd.plot(ax=ax2, color='white', linewidth=2, label="30-day SMA (YTD)")

# Label final YTD value, shifted right
last_date_ytd = vix_ytd.index[-1]
last_val_ytd = vix_ytd.iloc[-1]
ax2.text(last_date_ytd + pd.Timedelta(days=10), last_val_ytd,
         f"{last_val_ytd:.2f}", color='white', fontsize=9, fontweight='bold',
         ha='left', va='center')

# Watermark
ax2.text(0.01, 1.08, 'Pearl Gray Equity and Research', transform=ax2.transAxes,
         fontsize=10, color='white', alpha=0.8, fontweight='bold', ha='left', va='top')

# Style
ax2.set_title("VIX Index & 30-day SMA (YTD)", fontsize=14, fontweight='bold', color='white')
ax2.set_xlabel("Date", fontsize=12, color='white', fontweight='bold')
ax2.set_ylabel("VIX Level", fontsize=12, color='white', fontweight='bold')
ax2.legend(facecolor='black', edgecolor='white', labelcolor='white', fontsize=10)
ax2.grid(True, linestyle='--', color='white', alpha=0.2)
ax2.tick_params(colors='white')
for spine in ax2.spines.values():
    spine.set_color('white')

plt.tight_layout()
plt.show()












```
