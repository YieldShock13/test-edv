```python
import pandas as pd
import matplotlib.pyplot as plt
import requests

# FRED API key
api_key = "83cf5af9953ef78768ee3b3186db216b"

# Fetch FRED data
def fetch_fred_series(series_id, start="2016-01-01", end="2025-07-01"):
    url = "https://api.stlouisfed.org/fred/series/observations"
    params = {
        "series_id": series_id,
        "api_key": api_key,
        "file_type": "json",
        "observation_start": start,
        "observation_end": end,
        "frequency": "m"
    }
    r = requests.get(url, params=params)
    data = r.json()["observations"]
    df = pd.DataFrame(data)
    df["date"] = pd.to_datetime(df["date"])
    df.set_index("date", inplace=True)
    df = df["value"].replace(".", pd.NA).astype(float)
    return df

# Pull data
unemployment = fetch_fred_series("UNRATE")
cpi = fetch_fred_series("CPIAUCSL")
inflation = (cpi.pct_change(12) * 100).rename("Inflation Rate (%)")

# Combine
df = pd.concat([unemployment.rename("Unemployment Rate (%)"), inflation], axis=1).dropna()

# Plot
plt.style.use("default")
fig, ax = plt.subplots(figsize=(12, 6), facecolor='black')
fig.patch.set_facecolor('black')
ax.set_facecolor('black')

# Plot lines
ax.plot(df.index, df["Unemployment Rate (%)"], label="Unemployment Rate (%)", color='white', linewidth=2)
ax.plot(df.index, df["Inflation Rate (%)"], label="Inflation Rate (%)", color='gray', linewidth=2)

# Final value labels
last_idx = df.index[-1]
ax.text(last_idx, df["Unemployment Rate (%)"].iloc[-1],
        f'{df["Unemployment Rate (%)"].iloc[-1]:.2f}%',
        fontsize=9, color='white', fontweight='bold', va='bottom', ha='right')

ax.text(last_idx, df["Inflation Rate (%)"].iloc[-1],
        f'{df["Inflation Rate (%)"].iloc[-1]:.2f}%',
        fontsize=9, color='white', fontweight='bold', va='bottom', ha='right')

# Watermark (adjusted upward)
ax.text(0.01, 1.08, 'Pearl Gray Equity and Research', transform=ax.transAxes,
        fontsize=10, color='white', alpha=0.8, fontweight='bold', ha='left', va='top')

# Labels and style
ax.set_title("U.S. Unemployment and Inflation Rates (2016–2025)", color='white', fontsize=14, fontweight='bold')
ax.set_xlabel("Date", fontsize=12, color='white', fontweight='bold')
ax.set_ylabel("Rate (%)", fontsize=12, color='white', fontweight='bold')
ax.legend(facecolor='black', edgecolor='white', labelcolor='white', fontsize=10)

ax.grid(True, linestyle='--', color='white', alpha=0.2)
ax.tick_params(axis='x', colors='white', labelsize=9)
ax.tick_params(axis='y', colors='white', labelsize=9)

for spine in ax.spines.values():
    spine.set_color('white')

plt.tight_layout()
plt.show()












```
