```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    "Maturity": ["1M", "2M", "3M", "4M", "6M", "1Y", "2Y", "3Y", "5Y", "7Y", "10Y", "20Y", "30Y"],
    "Latest": [4.313, 4.358, 4.343, 4.350, 4.264, 4.071, 3.848, 3.805, 3.910, 4.128, 4.372, 4.938, 4.945],
    "1M ago": [4.150, 4.508, 4.305, 4.346, 4.277, 4.067, 3.904, 3.858, 3.959, 4.152, 4.376, 4.901, 4.889],
    "6M ago": [4.310, 4.324, 4.309, 4.333, 4.319, 4.201, 4.270, 4.315, 4.385, 4.476, 4.564, 4.872, 4.797],
    "12M ago": [5.369, 5.378, 5.342, 5.296, 5.201, 4.888, 4.517, 4.313, 4.170, 4.187, 4.222, 4.558, 4.451]
}

df = pd.DataFrame(data)

# Setup black background
plt.style.use('default')
fig, ax = plt.subplots(figsize=(14, 8))
fig.patch.set_facecolor('black')
ax.set_facecolor('black')

# Define lines with distinct styles and tags
lines = {
    'Latest':     {'color': 'white',     'lw': 2.5, 'ls': '-',  'label': 'Latest'},
    '1M ago':     {'color': 'lightgray', 'lw': 2,   'ls': '--', 'label': '1M ago'},
    '6M ago':     {'color': 'darkgray',  'lw': 2,   'ls': '--', 'label': '6M ago'},
    '12M ago':    {'color': 'gray',      'lw': 2,   'ls': '--', 'label': '12M ago'},
}

# Plot each line
for label, style in lines.items():
    ax.plot(df["Maturity"], df[label], marker='o', color=style['color'],
            linewidth=style['lw'], linestyle=style['ls'], label=style['label'])

    # Add end tag
    x = df["Maturity"].iloc[-1]
    y = df[label].iloc[-1]
    ax.annotate(label, (x, y), textcoords="offset points", xytext=(10, 0),
                ha='left', va='center', fontsize=9, color=style['color'])

# Add white data labels to 'Latest'
for i in range(len(df)):
    ax.annotate(f'{df["Latest"][i]:.2f}%',
                (df["Maturity"][i], df["Latest"][i]),
                textcoords="offset points",
                xytext=(0, 6),
                ha='center',
                fontsize=9,
                color='white')

# Title, labels, grid
ax.set_title("US Yield Curve Comparison", fontsize=14, color='white')
ax.set_xlabel("Maturity", fontsize=12, color='white')
ax.set_ylabel("Yield (%)", fontsize=12, color='white')
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')
ax.grid(True, linestyle='--', linewidth=0.5, color='gray')
ax.legend(facecolor='black', edgecolor='white', labelcolor='white')

# Watermark in white
fig.text(0.01, 0.985, "Pearl Gray Equity and Research", fontsize=10, color='white', alpha=0.9)

plt.tight_layout()
plt.show()








```
