```python
# === Code 1 (exact behavior) — now with US10Y included ===
# Upload:
#   "USEPUINDXM (2).csv"
#   "SPY Historical Data (2).csv"
#   "IEF Historical Data (1).csv"
#   "US Dollar Index Historical Data (2).csv"
#   "WTI_USD Historical Data (1).csv"
#   "XAU_USD Historical Data (2).csv"
#   "United States 10-Year Bond Yield Historical Data (5).csv"

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager as fm
from statsmodels.api import OLS, add_constant
from statsmodels.stats.stattools import durbin_watson
from pathlib import Path

# -------------------- CONFIG --------------------
EPU_CSV   = "USEPUINDXM (2).csv"

ASSET_CSVS = {
    "SPY":  "SPY Historical Data (2).csv",
    "IEF":  "IEF Historical Data (1).csv",
    "DXY":  "US Dollar Index Historical Data (2).csv",
    "WTI":  "WTI_USD Historical Data (1).csv",
    "Gold": "XAU_USD Historical Data (2).csv",
    "US10Y":"United States 10-Year Bond Yield Historical Data (5).csv",
}

START_A = "2007-01-01"
START_B = "2015-01-01"
START_C = "2024-08-01"

FONT_PATH  = "PearlGray.ttf"        # optional in original Code 1
BRAND_NAME = "Pearl Gray Equity and Research"

SAVE_FIGS  = True
OUT_DIR    = "figs_epu_code1_all_assets"
# -----------------------------------------------

# --- Theme (as in Code 1) ---
plt.rcParams.update({
    "figure.facecolor": "black",
    "axes.facecolor": "black",
    "savefig.facecolor": "black",
    "text.color": "white",
    "axes.labelcolor": "white",
    "axes.edgecolor": "white",
    "xtick.color": "white",
    "ytick.color": "white",
    "grid.color": "0.5",
    "grid.linestyle": "--",
    "grid.alpha": 0.3,
})

def set_brand_font(font_path: str):
    p = Path(font_path)
    if p.exists():
        fm.fontManager.addfont(str(p))
        font_name = fm.FontProperties(fname=str(p)).get_name()
        plt.rcParams["font.family"] = font_name
        return font_name
    else:
        print(f"[WARN] Font file not found: {font_path}. Using default Matplotlib font.")
        return None

_ = set_brand_font(FONT_PATH)

# --- Loader for EPU CSV (original) ---
def load_epu_series(csv_path: str) -> pd.Series:
    df = pd.read_csv(csv_path)
    dcols = [c for c in df.columns if "date" in c.lower()]
    dcol = dcols[0] if dcols else df.columns[0]
    df[dcol] = pd.to_datetime(df[dcol], errors="coerce")
    df = df.dropna(subset=[dcol])

    # Take first non-date column as value
    val_col = [c for c in df.columns if c != dcol][0]
    df[val_col] = pd.to_numeric(df[val_col], errors="coerce")
    df = df.dropna(subset=[val_col])

    df["YearMonth"] = df[dcol].dt.to_period("M")
    df = df.sort_values(dcol).reset_index(drop=True)
    return pd.Series(df[val_col].values, index=df["YearMonth"], name="USEPUINDXM")

# --- Loader for asset CSVs (original; no '%' stripping) ---
def load_asset_series(csv_path: str) -> pd.Series:
    df = pd.read_csv(csv_path)
    dcols = [c for c in df.columns if "date" in c.lower()]
    if not dcols:
        raise ValueError(f"No Date column found in {csv_path}")
    dcol = dcols[0]
    df[dcol] = pd.to_datetime(df[dcol], errors="coerce")
    df = df.dropna(subset=[dcol])

    # Identify price column
    candidates = ["Adj Close","Adjusted Close","Close","Price","Value","Index"]
    val_col = None
    for c in candidates:
        matches = [col for col in df.columns if col.lower() == c.lower()]
        if matches:
            val_col = matches[0]
            break
    if val_col is None:
        raise ValueError(f"No price/value column found in {csv_path}")

    # Minimal cleaning: commas and spaces only (as in Code 1)
    df[val_col] = pd.to_numeric(
        df[val_col].astype(str).str.replace(",","").str.replace(" ",""),
        errors="coerce"
    )
    df = df.dropna(subset=[val_col])

    df["YearMonth"] = df[dcol].dt.to_period("M")
    df = df.sort_values(dcol).reset_index(drop=True)
    return pd.Series(df[val_col].values, index=df["YearMonth"], name=Path(csv_path).stem)

# --- Load data ---
epu = load_epu_series(EPU_CSV)
assets = {name: load_asset_series(path) for name, path in ASSET_CSVS.items()}

print("\n=== Date Ranges Detected ===")
print(f"EPU: {epu.index.min()} → {epu.index.max()}")
for k,v in assets.items():
    print(f"{k}: {v.index.min()} → {v.index.max()}")

# --- Lag EPU by 1 month (LEVEL, not % change) ---
epu_lagged = epu.shift(1)  # <-- this is Code 1's independent variable

# --- Function: Run regressions (asset returns = m/m %) ---
def run_windows(asset_name, asset_series, epu_series):
    asset_ret = asset_series.pct_change()
    df = pd.DataFrame({
        "AssetRet": asset_ret,
        "EPU_lag": epu_series
    }).dropna()

    print(f"\n--- Alignment Preview: {asset_name} ---")
    print(df.head(5))
    print(df.tail(5))

    results = {}
    for label, start in [("Since_2007", START_A), ("Since_2015", START_B), ("Since_2024_08", START_C)]:
        w = df[df.index >= pd.Period(start, freq="M")]
        if len(w) < 3:
            results[label] = {"n": len(w), "r2": np.nan, "t_beta": np.nan, "dw": np.nan}
            continue
        X = add_constant(w["EPU_lag"].values)
        model = OLS(w["AssetRet"].values, X).fit()
        results[label] = {
            "n": len(w),
            "r2": model.rsquared,
            "t_beta": model.tvalues[1],
            "dw": durbin_watson(model.resid),
            "window_df": w,
            "model": model
        }
    return results

# --- Plot helper ---
def plot_asset_panels(asset_name, windows_results):
    titles = {
        "Since_2007": "2007–Present",
        "Since_2015": "2015–Present",
        "Since_2024_08": "Aug 2024–Present"
    }
    fig, axes = plt.subplots(1, 3, figsize=(16, 5), dpi=140, sharey=True)
    for ax, key in zip(axes, ["Since_2007", "Since_2015", "Since_2024_08"]):
        res = windows_results[key]
        ax.grid(True)
        if res["n"] and res["n"] > 2 and not np.isnan(res["r2"]):
            w = res["window_df"]
            ax.scatter(w["EPU_lag"], w["AssetRet"], s=18)
            beta0, beta1 = res["model"].params
            x_line = np.linspace(w["EPU_lag"].min(), w["EPU_lag"].max(), 100)
            y_line = beta0 + beta1 * x_line
            ax.plot(x_line, y_line, linewidth=1.8)
            ann = f"R²={res['r2']:.3f}\nt={res['t_beta']:.2f}\nDW={res['dw']:.2f}\nn={res['n']}"
        else:
            ann = f"Insufficient data\nn={res['n']}"
        ax.set_title(titles[key], fontsize=12, fontweight="bold")
        ax.set_xlabel("Lagged EPU (level, t-1)")  # original Code 1 labeling
        if key == "Since_2007":
            ax.set_ylabel("Asset Return (m/m)")
        ax.text(0.02, 0.98, ann, transform=ax.transAxes, va="top", ha="left", fontsize=10)
    fig.suptitle(f"{asset_name}: Return vs Lagged EPU (Level) — Three Windows", fontsize=16, fontweight="bold")
    fig.text(0.01, 0.98, BRAND_NAME, fontsize=10, color="0.7")
    fig.tight_layout(rect=[0, 0, 1, 0.95])
    if SAVE_FIGS:
        Path(OUT_DIR).mkdir(exist_ok=True)
        out = Path(OUT_DIR) / f"{asset_name}_lagged_epu_panels.png"
        plt.savefig(out, bbox_inches="tight")
        print(f"[Saved] {out}")
    plt.show()

# --- Run all assets with pure Code 1 logic ---
summary_rows = []
for name, s in assets.items():
    res = run_windows(name, s, epu_lagged)
    for k, v in res.items():
        summary_rows.append({
            "Asset": name, "Window": k,
            "n": v["n"], "R2": v["r2"], "t_beta": v["t_beta"], "DurbinWatson": v["dw"]
        })
    plot_asset_panels(name, res)

summary_df = pd.DataFrame(summary_rows).sort_values(["Asset", "Window"]).reset_index(drop=True)
print("\n=== Regression Summary ===")
print(summary_df.to_string(index=False))

Path(OUT_DIR).mkdir(exist_ok=True)
summary_df.to_csv(Path(OUT_DIR) / "epu_asset_regression_summary.csv", index=False)
print(f"[Saved] {Path(OUT_DIR) / 'epu_asset_regression_summary.csv'}")

print("\n=== Final Date Check ===")
print(f"EPU: {epu.index.min()} → {epu.index.max()}")
for k, v in assets.items():
    print(f"{k}: {v.index.min()} → {v.index.max()}")





















```
