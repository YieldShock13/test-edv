```python
# ============================================================
# GBPUSD OU PROJECT — STRICT SCREENSHOT FLOW
# ============================================================
# Flow:
# 1) Clean GBPUSD data properly
# 2) Define one state variable X_t
# 3) Compute next move X_{t+1} - X_t
# 4) Build nonparametric drift diagnostic
# 5) Build nonparametric diffusion diagnostic
# 6) Decide whether OU is a reasonable specification
# 7) Estimate OU via exact transition-density MLE
# 8) Compute half-life
# 9) Produce density outputs:
#       - conditional density
#       - unconditional density
#       - simulated histogram / kernel density
# 10) Produce first-passage outputs:
#       - hitting-time distribution
#       - target-hit probabilities over horizons
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.optimize import minimize
from scipy.stats import norm, gaussian_kde

# -----------------------------
# USER SETTINGS
# -----------------------------
FILE_NAME = "GBP_USD Historical Data (4).csv"

# One state variable only:
# X_t = log(price) - rolling mean of log(price)
ROLLING_WINDOW = 50

N_BUCKETS = 20
TRADING_DAYS_PER_YEAR = 252
DT = 1.0  # daily step

# Diagnostic thresholds for the OU decision step
LINEARITY_R2_THRESHOLD = 0.60
DIFFUSION_CV_THRESHOLD = 0.20

# Density / simulation settings
DENSITY_HORIZON_DAYS = 20
N_SIM_DENSITY = 20000

# First-passage settings
N_PATHS_FPT = 20000
MAX_FPT_HORIZON = 120
HORIZONS_TO_REPORT = [5, 10, 20, 30, 40, 50]

# Plot settings
plt.style.use("seaborn-v0_8-darkgrid")
FIG_WIDE = (13, 6)
FIG_STD = (9, 6)

# -----------------------------
# 1) LOAD & CLEAN DATA
# -----------------------------
df = pd.read_csv(FILE_NAME)
df.columns = [c.strip() for c in df.columns]

if "Date" not in df.columns or "Price" not in df.columns:
    raise ValueError("The file must contain 'Date' and 'Price' columns.")

df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
df["Price"] = (
    df["Price"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .str.replace("%", "", regex=False)
    .str.strip()
)
df["Price"] = pd.to_numeric(df["Price"], errors="coerce")

df = (
    df.dropna(subset=["Date", "Price"])
      .sort_values("Date")
      .drop_duplicates(subset="Date")
      .set_index("Date")
      .copy()
)

if len(df) < ROLLING_WINDOW + 30:
    raise ValueError("Not enough clean observations after loading the file.")

# -----------------------------
# 2) DEFINE ONE STATE VARIABLE X_t
# -----------------------------
df["log_price"] = np.log(df["Price"])
df["rolling_mean_log_price"] = df["log_price"].rolling(ROLLING_WINDOW).mean()
df["X"] = df["log_price"] - df["rolling_mean_log_price"]

df = df.dropna(subset=["X"]).copy()

# -----------------------------
# 3) COMPUTE NEXT MOVE X_{t+1} - X_t
# -----------------------------
df["dX_next"] = df["X"].shift(-1) - df["X"]
df = df.dropna(subset=["dX_next"]).copy()

if df["X"].nunique() < 10:
    raise ValueError("State variable has too little variation for this workflow.")

# -----------------------------
# HELPER FUNCTIONS
# -----------------------------
def build_buckets(series, n_buckets):
    n_use = min(n_buckets, max(5, series.nunique() // 2))
    buckets = pd.qcut(series, q=n_use, duplicates="drop")
    return buckets

def linear_fit_r2(x, y):
    x = np.asarray(x)
    y = np.asarray(y)
    coef = np.polyfit(x, y, 1)
    y_hat = np.polyval(coef, x)
    ss_res = np.sum((y - y_hat) ** 2)
    ss_tot = np.sum((y - np.mean(y)) ** 2)
    r2 = np.nan if ss_tot == 0 else 1 - ss_res / ss_tot
    return coef, y_hat, r2

def ou_neg_loglik(params, x, dt=1.0):
    kappa, mu, sigma = params
    if kappa <= 0 or sigma <= 0:
        return 1e12

    x_t = x[:-1]
    x_tp1 = x[1:]

    exp_term = np.exp(-kappa * dt)
    cond_mean = mu + (x_t - mu) * exp_term
    cond_var = (sigma**2 / (2 * kappa)) * (1 - np.exp(-2 * kappa * dt))

    if np.any(cond_var <= 0) or np.isnan(cond_var).any():
        return 1e12

    ll = norm.logpdf(x_tp1, loc=cond_mean, scale=np.sqrt(cond_var)).sum()
    return -ll

def fit_ou_mle(x, dt=1.0):
    x = np.asarray(x, dtype=float)

    # Simple initial values from the state itself
    mu0 = np.mean(x)
    sigma0 = np.std(np.diff(x), ddof=1)
    sigma0 = max(sigma0, 1e-4)
    kappa0 = 0.05

    init = np.array([kappa0, mu0, sigma0])

    bounds = [
        (1e-6, 5.0),                                  # kappa
        (np.min(x) - 5*np.std(x), np.max(x) + 5*np.std(x)),  # mu
        (1e-6, 5*np.std(x) + 1e-6)                   # sigma
    ]

    res = minimize(
        ou_neg_loglik,
        x0=init,
        args=(x, dt),
        method="L-BFGS-B",
        bounds=bounds
    )

    if not res.success:
        raise RuntimeError(f"OU MLE failed: {res.message}")

    kappa, mu, sigma = res.x
    return kappa, mu, sigma, res

def ou_conditional_mean_var(x0, kappa, mu, sigma, horizon, dt=1.0):
    t = horizon * dt
    mean_t = mu + (x0 - mu) * np.exp(-kappa * t)
    var_t = (sigma**2 / (2 * kappa)) * (1 - np.exp(-2 * kappa * t))
    return mean_t, var_t

def simulate_ou_terminal_values(x0, kappa, mu, sigma, horizon, n_sims, dt=1.0, seed=42):
    rng = np.random.default_rng(seed)
    mean_t, var_t = ou_conditional_mean_var(x0, kappa, mu, sigma, horizon, dt)
    return rng.normal(loc=mean_t, scale=np.sqrt(var_t), size=n_sims)

def simulate_ou_paths(x0, kappa, mu, sigma, horizon, n_paths, dt=1.0, seed=42):
    rng = np.random.default_rng(seed)
    paths = np.zeros((n_paths, horizon + 1))
    paths[:, 0] = x0

    exp_term = np.exp(-kappa * dt)
    step_var = (sigma**2 / (2 * kappa)) * (1 - np.exp(-2 * kappa * dt))
    step_sd = np.sqrt(step_var)

    for t in range(1, horizon + 1):
        z = rng.standard_normal(n_paths)
        paths[:, t] = mu + (paths[:, t-1] - mu) * exp_term + step_sd * z

    return paths

def first_hit_times(paths, level, direction):
    n_paths, horizon_plus_one = paths.shape
    horizon = horizon_plus_one - 1
    hits = np.zeros(n_paths, dtype=bool)
    times = np.full(n_paths, np.nan)

    for i in range(n_paths):
        series = paths[i, 1:]  # exclude t=0 for first-passage
        if direction == "down":
            idx = np.where(series <= level)[0]
        elif direction == "up":
            idx = np.where(series >= level)[0]
        else:
            raise ValueError("direction must be 'down' or 'up'")

        if len(idx) > 0:
            hits[i] = True
            times[i] = idx[0] + 1

    return hits, times

# -----------------------------
# 4) NONPARAMETRIC DRIFT DIAGNOSTIC
# -----------------------------
df["bucket"] = build_buckets(df["X"], N_BUCKETS)

drift_diag = (
    df.groupby("bucket", observed=True)
      .agg(X_mean=("X", "mean"),
           dX_mean=("dX_next", "mean"),
           count=("dX_next", "size"))
      .dropna()
      .reset_index(drop=True)
)

if len(drift_diag) < 5:
    raise ValueError("Too few usable buckets for the nonparametric drift diagnostic.")

drift_coef, drift_line, drift_r2 = linear_fit_r2(drift_diag["X_mean"], drift_diag["dX_mean"])

# -----------------------------
# 5) NONPARAMETRIC DIFFUSION DIAGNOSTIC
# -----------------------------
diff_diag = (
    df.groupby("bucket", observed=True)
      .agg(X_mean=("X", "mean"),
           dX_std=("dX_next", "std"),
           count=("dX_next", "size"))
      .dropna()
      .reset_index(drop=True)
)

if len(diff_diag) < 5:
    raise ValueError("Too few usable buckets for the nonparametric diffusion diagnostic.")

diff_mean = diff_diag["dX_std"].mean()
diff_cv = diff_diag["dX_std"].std(ddof=1) / diff_mean if diff_mean > 0 else np.nan

# -----------------------------
# 6) DECIDE WHETHER OU IS THE RIGHT SPECIFICATION
# -----------------------------
left_mask = drift_diag["X_mean"] < 0
right_mask = drift_diag["X_mean"] > 0

left_avg = drift_diag.loc[left_mask, "dX_mean"].mean() if left_mask.any() else np.nan
right_avg = drift_diag.loc[right_mask, "dX_mean"].mean() if right_mask.any() else np.nan

drift_shape_ok = (left_avg > 0) and (right_avg < 0)
linearity_ok = np.isfinite(drift_r2) and (drift_r2 >= LINEARITY_R2_THRESHOLD)
diffusion_ok = np.isfinite(diff_cv) and (diff_cv <= DIFFUSION_CV_THRESHOLD)

ou_decision = drift_shape_ok and linearity_ok and diffusion_ok

# -----------------------------
# 7) ESTIMATE OU VIA TRANSITION-DENSITY MLE
# -----------------------------
x = df["X"].values
kappa, mu, sigma, mle_result = fit_ou_mle(x, dt=DT)

# -----------------------------
# 8) COMPUTE HALF-LIFE
# -----------------------------
half_life = np.log(2) / kappa

# -----------------------------
# 9) DENSITY OUTPUTS
#    - conditional density
#    - unconditional density
#    - simulated histogram / kernel density
# -----------------------------
x0 = df["X"].iloc[-1]

# Conditional density at chosen horizon
cond_mean, cond_var = ou_conditional_mean_var(
    x0=x0, kappa=kappa, mu=mu, sigma=sigma, horizon=DENSITY_HORIZON_DAYS, dt=DT
)
cond_sd = np.sqrt(cond_var)

# Unconditional (stationary) density
uncond_var = sigma**2 / (2 * kappa)
uncond_sd = np.sqrt(uncond_var)

# Simulated terminal values + KDE
terminal_vals = simulate_ou_terminal_values(
    x0=x0,
    kappa=kappa,
    mu=mu,
    sigma=sigma,
    horizon=DENSITY_HORIZON_DAYS,
    n_sims=N_SIM_DENSITY,
    dt=DT,
    seed=123
)
terminal_kde = gaussian_kde(terminal_vals)

# -----------------------------
# 10) FIRST-PASSAGE OUTPUTS
#     - hitting-time distribution
#     - target-hit probabilities over horizons
# -----------------------------
# Target level: fitted equilibrium mu
# Direction depends on whether current state is above or below mu
target_level = mu
direction = "down" if x0 > mu else "up"

paths = simulate_ou_paths(
    x0=x0,
    kappa=kappa,
    mu=mu,
    sigma=sigma,
    horizon=MAX_FPT_HORIZON,
    n_paths=N_PATHS_FPT,
    dt=DT,
    seed=999
)

hit_flags, hit_times = first_hit_times(paths, level=target_level, direction=direction)

hit_prob_by_horizon = {}
for h in HORIZONS_TO_REPORT:
    hit_prob_by_horizon[h] = np.mean(hit_flags & (hit_times <= h))

valid_hit_times = hit_times[~np.isnan(hit_times)]

# -----------------------------
# PLOTS
# -----------------------------

# Raw price
fig, ax = plt.subplots(figsize=FIG_WIDE)
ax.plot(df.index, df["Price"], linewidth=1.6)
ax.set_title("GBPUSD Daily Price")
ax.set_ylabel("Price")
plt.tight_layout()
plt.show()

# State variable
fig, ax = plt.subplots(figsize=FIG_WIDE)
ax.plot(df.index, df["X"], linewidth=1.4, label="State variable X")
ax.axhline(0, linestyle="--", linewidth=1.0)
ax.set_title(f"State Variable: log(price) - {ROLLING_WINDOW}-day rolling mean of log(price)")
ax.set_ylabel("X")
ax.legend()
plt.tight_layout()
plt.show()

# Drift diagnostic
fig, ax = plt.subplots(figsize=FIG_STD)
ax.scatter(drift_diag["X_mean"], drift_diag["dX_mean"], s=55, alpha=0.9)
ax.plot(drift_diag["X_mean"], drift_line, linewidth=1.8, linestyle="--")
ax.axhline(0, linestyle="--", linewidth=1.0)
ax.axvline(0, linestyle="--", linewidth=1.0)
ax.set_title("Nonparametric Drift Diagnostic")
ax.set_xlabel("Bucket mean of X")
ax.set_ylabel("Average next move: X(t+1) - X(t)")
plt.tight_layout()
plt.show()

# Diffusion diagnostic
fig, ax = plt.subplots(figsize=FIG_STD)
ax.scatter(diff_diag["X_mean"], diff_diag["dX_std"], s=55, alpha=0.9)
ax.axvline(0, linestyle="--", linewidth=1.0)
ax.set_title("Nonparametric Diffusion Diagnostic")
ax.set_xlabel("Bucket mean of X")
ax.set_ylabel("Std. dev. of next move")
plt.tight_layout()
plt.show()

# Conditional + unconditional densities
x_grid_min = min(cond_mean - 4*cond_sd, mu - 4*uncond_sd, terminal_vals.min())
x_grid_max = max(cond_mean + 4*cond_sd, mu + 4*uncond_sd, terminal_vals.max())
x_grid = np.linspace(x_grid_min, x_grid_max, 800)

fig, ax = plt.subplots(figsize=FIG_STD)
ax.plot(x_grid, norm.pdf(x_grid, cond_mean, cond_sd), linewidth=2.0, label=f"Conditional density ({DENSITY_HORIZON_DAYS}d)")
ax.plot(x_grid, norm.pdf(x_grid, mu, uncond_sd), linewidth=2.0, linestyle="--", label="Unconditional density")
ax.axvline(x0, linewidth=1.2, linestyle=":", label="Current X")
ax.axvline(mu, linewidth=1.2, linestyle="-.", label="Fitted μ")
ax.set_title("Conditional and Unconditional Densities")
ax.set_xlabel("State")
ax.set_ylabel("Density")
ax.legend()
plt.tight_layout()
plt.show()

# Simulated histogram + KDE of terminal state
fig, ax = plt.subplots(figsize=FIG_STD)
ax.hist(terminal_vals, bins=50, density=True, alpha=0.45, label="Simulated histogram")
ax.plot(x_grid, terminal_kde(x_grid), linewidth=2.0, label="Kernel density")
ax.plot(x_grid, norm.pdf(x_grid, cond_mean, cond_sd), linewidth=1.8, linestyle="--", label="Analytical conditional density")
ax.set_title(f"Simulated Terminal-State Density ({DENSITY_HORIZON_DAYS} Days)")
ax.set_xlabel("State at horizon")
ax.set_ylabel("Density")
ax.legend()
plt.tight_layout()
plt.show()

# First-passage histogram + KDE
if len(valid_hit_times) > 10:
    fpt_grid = np.linspace(1, MAX_FPT_HORIZON, 500)
    fpt_kde = gaussian_kde(valid_hit_times)

    fig, ax = plt.subplots(figsize=FIG_STD)
    ax.hist(valid_hit_times, bins=min(40, MAX_FPT_HORIZON), density=True, alpha=0.45, label="Hitting-time histogram")
    ax.plot(fpt_grid, fpt_kde(fpt_grid), linewidth=2.0, label="Kernel density")
    ax.set_title("First-Passage Time Distribution to Target Level")
    ax.set_xlabel("Days to first hit")
    ax.set_ylabel("Density")
    ax.legend()
    plt.tight_layout()
    plt.show()

# -----------------------------
# OUTPUTS
# -----------------------------
print("\n===== STEP 6: OU SPECIFICATION DECISION =====")
print(f"Drift shape condition met: {drift_shape_ok}")
print(f"Drift linearity R^2: {drift_r2:.4f}")
print(f"Diffusion coefficient of variation: {diff_cv:.4f}")
print(f"OU decision (roughly linear drift + roughly constant diffusion): {ou_decision}")

print("\n===== STEP 7: OU MLE =====")
print(f"kappa: {kappa:.6f}")
print(f"mu   : {mu:.6f}")
print(f"sigma: {sigma:.6f}")

print("\n===== STEP 8: HALF-LIFE =====")
print(f"half-life: {half_life:.2f} trading days")

print("\n===== STEP 9: DENSITY OUTPUTS =====")
print(f"Current state X0: {x0:.6f}")
print(f"{DENSITY_HORIZON_DAYS}-day conditional mean: {cond_mean:.6f}")
print(f"{DENSITY_HORIZON_DAYS}-day conditional sd  : {cond_sd:.6f}")
print(f"Unconditional mean: {mu:.6f}")
print(f"Unconditional sd  : {uncond_sd:.6f}")

print("\n===== STEP 10: FIRST-PASSAGE OUTPUTS =====")
print(f"Target level: {target_level:.6f}")
print(f"Direction to target: {direction}")
for h in HORIZONS_TO_REPORT:
    print(f"Probability of hitting target within {h:>2} days: {hit_prob_by_horizon[h]:.2%}")

if len(valid_hit_times) > 0:
    print(f"Mean hitting time (conditional on hit): {np.mean(valid_hit_times):.2f} days")
    print(f"Median hitting time (conditional on hit): {np.median(valid_hit_times):.2f} days")
else:
    print("No simulated hits reached the target within the maximum horizon.")

    # ============================================================
# TRADE LAYER — ENTRY / TARGET / FIRST-PASSAGE
# ============================================================

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# ============================================================
# REQUIREMENTS (from previous code)
# ============================================================
# must already exist:
# mu, sigma, kappa
# (from OU estimation)

# ============================================================
# 1) DEFINE LEVELS
# ============================================================

entry_sigma = 1.0
target_sigma = 0.5
stop_sigma = 2.0  # optional

entry_level_up = mu + entry_sigma * sigma
entry_level_down = mu - entry_sigma * sigma

target_level_up = mu + target_sigma * sigma
target_level_down = mu - target_sigma * sigma

stop_level_up = mu + stop_sigma * sigma
stop_level_down = mu - stop_sigma * sigma

# ============================================================
# 2) SIMULATION FUNCTION (OU)
# ============================================================

def simulate_paths(x0, kappa, mu, sigma, n_paths=20000, horizon=60):
    paths = np.zeros((n_paths, horizon))
    paths[:, 0] = x0

    dt = 1
    exp_term = np.exp(-kappa * dt)
    var = (sigma**2 / (2 * kappa)) * (1 - np.exp(-2 * kappa * dt))
    sd = np.sqrt(var)

    for t in range(1, horizon):
        z = np.random.normal(size=n_paths)
        paths[:, t] = mu + (paths[:, t-1] - mu)*exp_term + sd*z

    return paths

# ============================================================
# 3) FIRST-PASSAGE FUNCTION
# ============================================================

def first_passage(paths, target, direction="down"):
    n_paths, T = paths.shape

    hit = np.zeros(n_paths, dtype=bool)
    hit_time = np.full(n_paths, np.nan)

    for i in range(n_paths):
        for t in range(1, T):
            if direction == "down" and paths[i,t] <= target:
                hit[i] = True
                hit_time[i] = t
                break
            elif direction == "up" and paths[i,t] >= target:
                hit[i] = True
                hit_time[i] = t
                break

    return hit, hit_time

# ============================================================
# 4) RUN TRADE SCENARIO (UPPER ENTRY)
# ============================================================

np.random.seed(42)

# start from +1σ
x0 = entry_level_up

paths = simulate_paths(x0, kappa, mu, sigma)

hit, hit_time = first_passage(paths, target_level_up, direction="down")

# ============================================================
# 5) FIRST-PASSAGE TIME DENSITY
# ============================================================

valid_times = hit_time[~np.isnan(hit_time)]

plt.figure(figsize=(8,6))
plt.hist(valid_times, bins=40, density=True, alpha=0.5)

if len(valid_times) > 10:
    kde = gaussian_kde(valid_times)
    x = np.linspace(1, 60, 200)
    plt.plot(x, kde(x))

plt.title("First-Passage Time Density (1σ → 0.5σ)")
plt.xlabel("Days")
plt.ylabel("Density")
plt.show()

# ============================================================
# 6) PROBABILITY VS TIME
# ============================================================

horizons = [5,10,15,20,30,40,50]

probs = []

for h in horizons:
    probs.append(np.mean(hit & (hit_time <= h)))

plt.figure(figsize=(8,6))
plt.plot(horizons, probs, marker="o")
plt.title("Probability of Hitting Target vs Time")
plt.xlabel("Days")
plt.ylabel("Probability")
plt.show()

# ============================================================
# 7) OUTPUT
# ============================================================

print("\n===== TRADE RESULTS =====")
print(f"Entry: +1σ = {entry_level_up:.6f}")
print(f"Target: +0.5σ = {target_level_up:.6f}")
print(f"Stop: +2σ = {stop_level_up:.6f}")

print("\nProbabilities:")
for h, p in zip(horizons, probs):
    print(f"{h} days: {p:.2%}")

if len(valid_times) > 0:
    print(f"\nMean time to target: {np.mean(valid_times):.2f} days")

    # ============================================================
# GBPUSD TRAJECTORY WITH PERCENTILE BANDS (NO Z-SCORE)
# ============================================================

import numpy as np
import matplotlib.pyplot as plt

# use your existing state variable X
# (df["X"] already exists)

X_series = df["X"].copy()

# rolling percentiles
window = 252  # 1 year rolling

p10 = X_series.rolling(window).quantile(0.10)
p25 = X_series.rolling(window).quantile(0.25)
p50 = X_series.rolling(window).quantile(0.50)
p75 = X_series.rolling(window).quantile(0.75)
p90 = X_series.rolling(window).quantile(0.90)

# ============================================================
# PLOT
# ============================================================

plt.figure(figsize=(12,6))

plt.plot(X_series, label="State X", linewidth=1.2)

plt.plot(p50, linestyle="--", linewidth=1.2, label="Median")

plt.fill_between(X_series.index, p25, p75, alpha=0.2, label="25–75%")
plt.fill_between(X_series.index, p10, p90, alpha=0.1, label="10–90%")

plt.axhline(0, linestyle="--", linewidth=1)

plt.title("GBPUSD State Trajectory with Rolling Percentile Bands")
plt.ylabel("X")
plt.legend()

plt.show()

# ============================================================
# SINGLE-FACTOR OLS: GBPUSD vs VIX + YIELD SPREAD
# ============================================================

import pandas as pd
import numpy as np
import statsmodels.api as sm

# -----------------------------
# LOAD DATA
# -----------------------------
gbp = df.copy()  # your existing GBPUSD dataframe

vix = pd.read_csv("CBOE Volatility Index Historical Data (12).csv")
spread = pd.read_csv("UK 10 Year vs US 10 Year Spread Bond Yield Historical Data.csv")

# -----------------------------
# CLEAN VIX
# -----------------------------
vix["Date"] = pd.to_datetime(vix["Date"])
vix = vix.sort_values("Date")
vix["Price"] = vix["Price"].astype(str).str.replace(",", "").astype(float)
vix = vix.set_index("Date")[["Price"]]
vix.rename(columns={"Price": "VIX"}, inplace=True)

# -----------------------------
# CLEAN SPREAD
# -----------------------------
spread["Date"] = pd.to_datetime(spread["Date"])
spread = spread.sort_values("Date")
spread["Price"] = spread["Price"].astype(str).str.replace(",", "").astype(float)
spread = spread.set_index("Date")[["Price"]]
spread.rename(columns={"Price": "Spread"}, inplace=True)

# -----------------------------
# MERGE
# -----------------------------
data = gbp[["price"]].copy()
data = data.join(vix, how="inner")
data = data.join(spread, how="inner")

# -----------------------------
# RETURNS (NO SMOOTHING)
# -----------------------------
data["ret"] = np.log(data["price"]).diff()

data = data.dropna()

# ============================================================
# 1) GBPUSD vs VIX
# ============================================================

X_vix = sm.add_constant(data["VIX"])
y = data["ret"]

model_vix = sm.OLS(y, X_vix).fit()

print("\n===== GBPUSD vs VIX =====")
print(model_vix.summary())

# ============================================================
# 2) GBPUSD vs YIELD SPREAD
# ============================================================

X_spread = sm.add_constant(data["Spread"])

model_spread = sm.OLS(y, X_spread).fit()

print("\n===== GBPUSD vs YIELD SPREAD =====")
print(model_spread.summary())
```
