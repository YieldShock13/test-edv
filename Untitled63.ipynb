```python
# === 10Y2Y (T10Y2Y) Daily: Raw Series + Z-Score ===
# Colab-ready. Uses your FRED CSV (T10Y2Y), plots daily series as-is, plus z-score.
# End date clipped to 2025-08-27. Watermark: Orcas Spread.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ---------- CONFIG ----------
T102_CSV_PATH   = "T10Y2Y (5).csv"   # <- update if filename differs
TARGET_END_DATE = "2025-08-27"

SERIES_NAME     = "10Y2Y"
Z_NAME          = "Z-Score (10Y2Y)"
WATERMARK_TEXT  = "Orcas Spread"

# Z-score: None = full-sample; or set an int for rolling window (in DAYS)
ROLLING_WINDOW  = None  # e.g., 756 for ~3Y of trading days

DPI = 160
FONT_FAMILY = "DejaVu Sans"
plt.rcParams.update({
    "font.family": FONT_FAMILY,
    "figure.facecolor": "black",
    "axes.facecolor": "black",
    "savefig.facecolor": "black",
    "text.color": "white",
    "axes.labelcolor": "white",
    "xtick.color": "white",
    "ytick.color": "white",
    "axes.grid": True,
    "grid.color": "#555555",
    "grid.alpha": 0.25
})
# ---------------------------

def read_series_to_bps(path):
    """Read CSV, auto-detect date/value columns, normalize to bps."""
    df = pd.read_csv(path)
    # detect date col
    date_col = None
    for c in df.columns:
        try:
            pd.to_datetime(df[c])
            date_col = c
            break
        except Exception:
            continue
    if date_col is None:
        raise ValueError(f"No parseable date column in {path}")

    # detect numeric col (use rightmost numeric)
    numeric_cols = [c for c in df.columns if pd.api.types.is_numeric_dtype(df[c]) and c != date_col]
    if not numeric_cols:
        for c in df.columns:
            if c != date_col:
                df[c] = pd.to_numeric(df[c].astype(str).str.replace(",",""), errors="coerce")
        numeric_cols = [c for c in df.columns if pd.api.types.is_numeric_dtype(df[c]) and c != date_col]
        if not numeric_cols:
            raise ValueError(f"No numeric value column in {path}")
    val_col = numeric_cols[-1]

    s = pd.Series(pd.to_numeric(df[val_col], errors="coerce").values,
                  index=pd.to_datetime(df[date_col]),
                  name=val_col).dropna().sort_index()

    # normalize to bps
    p95 = float(s.quantile(0.95))
    if p95 <= 1.0:        # fraction -> bps (e.g., 0.0123)
        s = s * 10000.0
    elif p95 <= 20.0:     # percent -> bps (e.g., 1.23)
        s = s * 100.0
    # else already bps
    return s

def zscore(series: pd.Series, rolling_window=None) -> pd.Series:
    if rolling_window is None:
        mu = series.mean()
        sd = series.std(ddof=1)
        return (series - mu) / sd
    mu = series.rolling(rolling_window, min_periods=max(30, rolling_window//3)).mean()
    sd = series.rolling(rolling_window, min_periods=max(30, rolling_window//3)).std(ddof=1)
    return (series - mu) / sd

def add_watermark(fig, text=WATERMARK_TEXT, alpha=0.25):
    fig.text(0.995, 0.02, text, fontsize=10, color="white",
             ha="right", va="bottom", alpha=alpha)

# --- Load daily series (as-is) ---
t102_daily = read_series_to_bps(T102_CSV_PATH).rename(SERIES_NAME)
t102_daily = t102_daily.loc[:TARGET_END_DATE].dropna()

# --- Z-score (daily) ---
z = zscore(t102_daily, rolling_window=ROLLING_WINDOW).rename(Z_NAME)

# --- Plot 1: Raw daily series (as-is) ---
fig1, ax1 = plt.subplots(figsize=(11,4), dpi=DPI)
ax1.plot(t102_daily.index, t102_daily.values, linewidth=2)
ax1.set_title("10Y2Y Spread (Daily, as-is)", pad=12)
ax1.set_ylabel("Basis Points")
ax1.set_xlabel("Date")
ax1.axhline(0, linestyle="--", alpha=0.5)
last_dt, last_val = t102_daily.index[-1], t102_daily.iloc[-1]
ax1.annotate(f"{last_val:.0f} bps ({last_dt.date()})",
             xy=(last_dt, last_val), xytext=(10,0),
             textcoords="offset points", va="center")
add_watermark(fig1)
plt.tight_layout()
plt.show()

# --- Plot 2: Z-Score (daily) ---
fig2, ax2 = plt.subplots(figsize=(11,4), dpi=DPI)
ax2.plot(z.index, z.values, linewidth=2)
ax2.set_title("10Y2Y Spread  Z-Score (Daily)", pad=12)
ax2.set_ylabel("Z")
ax2.set_xlabel("Date")
ax2.axhline(0, linestyle="--", alpha=0.5)
for thr, ls in [(1,"-."),(2,":")]:
    ax2.axhline(thr, linestyle=ls, alpha=0.6)
    ax2.axhline(-thr, linestyle=ls, alpha=0.6)
ax2.fill_between(z.index, 1, z, where=z>=1, alpha=0.12)
ax2.fill_between(z.index, 2, z, where=z>=2, alpha=0.18)
ax2.fill_between(z.index, -1, z, where=z<=-1, alpha=0.12)
ax2.fill_between(z.index, -2, z, where=z<=-2, alpha=0.18)
ax2.annotate(f"{z.iloc[-1]:.2f}  ({z.index[-1].date()})",
             xy=(z.index[-1], z.iloc[-1]),
             xytext=(10,0), textcoords="offset points", va="center")
add_watermark(fig2)
plt.tight_layout()
plt.show()

# --- Save outputs ---
t102_daily.to_frame().to_csv("t10y2y_daily_bps.csv")
z.to_frame().to_csv("t10y2y_daily_zscore.csv")
```
