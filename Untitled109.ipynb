```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

pd.options.display.float_format = "{:,.4f}".format

# -------------------------------------------------------
# CONFIG
# -------------------------------------------------------
BASE_CAPITAL = 1_000_000.0        # only used for equity curves
START_DATE   = "2020-01-01"       # regression starts here

SPX_CSV   = "/content/S&P 500 Historical Data (6).csv"
VIX_CSV   = "/content/CBOE Volatility Index Historical Data (4).csv"
YIELD_CSV = "/content/United States 1-Month Bond Yield Historical Data.csv"

DATE_COL  = "Date"
PRICE_COL = "Price"

T              = 1.0 / 12.0   # 1-month option
DIVIDEND_YIELD = 0.02         # SPX dividend yield proxy
MIN_BETA_WIN   = 12           # min months for rolling beta


# -------------------------------------------------------
# LOAD DATA
# -------------------------------------------------------
# SPX
spx = pd.read_csv(SPX_CSV)
spx[DATE_COL]  = pd.to_datetime(spx[DATE_COL], format="%Y/%m/%d")
spx[PRICE_COL] = spx[PRICE_COL].astype(str).str.replace(",", "", regex=False).astype(float)
spx            = spx.sort_values(DATE_COL).rename(columns={PRICE_COL: "spx_price"})

# VIX
vix = pd.read_csv(VIX_CSV)
vix[DATE_COL]  = pd.to_datetime(vix[DATE_COL], format="%Y/%m/%d")
vix[PRICE_COL] = vix[PRICE_COL].astype(str).str.replace(",", "", regex=False).astype(float)
vix            = vix.sort_values(DATE_COL).rename(columns={PRICE_COL: "vix_price"})

# 1M yield (investing.com, price = yield % p.a.)
yld = pd.read_csv(YIELD_CSV)
yld[DATE_COL]  = pd.to_datetime(yld[DATE_COL], format="%Y/%m/%d")
yld[PRICE_COL] = yld[PRICE_COL].astype(str).str.replace(",", "", regex=False).astype(float)
yld            = yld.sort_values(DATE_COL).rename(columns={PRICE_COL: "rf_yield_pct"})

# Merge
df = (
    spx.merge(vix, on=DATE_COL, how="inner")
       .merge(yld, on=DATE_COL, how="inner")
       .sort_values(DATE_COL)
       .reset_index(drop=True)
)
df = df[df[DATE_COL] >= START_DATE].reset_index(drop=True)

# Basic series
df["spx_ret"]      = df["spx_price"].pct_change()
df["rf_yield_ann"] = df["rf_yield_pct"] / 100.0
df["rf_month"]     = df["rf_yield_ann"] / 12.0
df["sigma_annual"] = df["vix_price"] / 100.0
df["r_cc"]         = np.log(1.0 + df["rf_yield_ann"])

df = df.dropna(subset=["spx_ret", "rf_month", "sigma_annual", "r_cc"]).reset_index(drop=True)
n  = len(df)


# -------------------------------------------------------
# BSM CALL
# -------------------------------------------------------
def norm_cdf(x: float) -> float:
    return 0.5 * (1.0 + math.erf(x / math.sqrt(2.0)))

def bsm_call_price(S: float, K: float, r_cc: float, sigma: float, T: float, q: float = 0.0) -> float:
    if T <= 0.0 or sigma <= 0.0:
        return max(S - K, 0.0)
    num = math.log(S / K) + (r_cc - q + 0.5 * sigma ** 2) * T
    den = sigma * math.sqrt(T)
    d1  = num / den
    d2  = d1 - sigma * math.sqrt(T)
    return S * math.exp(-q * T) * norm_cdf(d1) - K * math.exp(-r_cc * T) * norm_cdf(d2)


# -------------------------------------------------------
# OPTIONS OVERLAY PORTFOLIO
# 100% in RF; each month use that month's interest as premium
# -------------------------------------------------------
df["opt_value"]     = np.nan
df["call_budget"]   = np.nan
df["call_price"]    = np.nan
df["call_cost_pct"] = np.nan
df["call_units"]    = np.nan
df["call_notional"] = np.nan
df["call_payoff"]   = 0.0
df["opt_particip"]  = np.nan
df["opt_trade_pnl"] = 0.0

df.loc[0, "opt_value"] = BASE_CAPITAL

for t in range(0, n - 1):
    P_t     = df.loc[t, "opt_value"]
    S_t     = df.loc[t, "spx_price"]
    sigma_t = df.loc[t, "sigma_annual"]
    r_cc_t  = df.loc[t, "r_cc"]
    rf_m_t  = df.loc[t, "rf_month"]

    # Interest for this month (budget), floored at 0
    I_t = max(P_t * rf_m_t, 0.0)

    # Price ATM 1M call
    K_t = S_t
    C_t = bsm_call_price(S=S_t, K=K_t, r_cc=r_cc_t, sigma=sigma_t,
                         T=T, q=DIVIDEND_YIELD)
    cost_pct_t = C_t / S_t if S_t > 0 else np.nan

    q_t       = 0.0
    notional = 0.0
    premium  = 0.0

    if C_t > 0 and I_t > 0:
        q_t      = I_t / C_t
        premium  = I_t
        notional = q_t * S_t

    # End-of-month
    S_next   = df.loc[t + 1, "spx_price"]
    payoff_t = q_t * max(S_next - K_t, 0.0)

    P_after_rf = P_t * (1.0 + rf_m_t)
    P_next     = P_after_rf - premium + payoff_t

    # Store
    df.loc[t, "call_budget"]     = premium
    df.loc[t, "call_price"]      = C_t
    df.loc[t, "call_cost_pct"]   = cost_pct_t
    df.loc[t, "call_units"]      = q_t
    df.loc[t, "call_notional"]   = notional
    df.loc[t + 1, "call_payoff"] = payoff_t
    df.loc[t + 1, "opt_value"]   = P_next

    df.loc[t, "opt_particip"]  = (notional / P_t) if P_t != 0 else np.nan
    df.loc[t, "opt_trade_pnl"] = payoff_t - premium

df["opt_particip"] = df["opt_particip"].ffill()

print("\n===== Head / Tail (Overlay Portfolio) =====")
print(df[[DATE_COL, "spx_price", "opt_value"]].head())
print(df[[DATE_COL, "spx_price", "opt_value"]].tail())


# -------------------------------------------------------
# RETURNS / ROLLING BETA / BETA-MATCHED BENCHMARK
# -------------------------------------------------------
df["opt_ret"] = df["opt_value"].pct_change()
df["mkt_ret"] = df["spx_price"].pct_change()
df["rf_ret"]  = df["rf_month"]  # monthly RF

# Rolling beta of overlay vs SPX
rolling_beta = pd.Series(index=df.index, dtype=float)

for i in range(n):
    valid = df.loc[:i, ["opt_ret", "mkt_ret"]].dropna()
    k = len(valid)
    if k >= MIN_BETA_WIN:
        x = valid["mkt_ret"]
        y = valid["opt_ret"]
        cov = np.cov(y, x)[0, 1]
        var = np.var(x)
        beta_i = cov / var if var != 0 else np.nan
    else:
        beta_i = np.nan
    rolling_beta.iloc[i] = beta_i

# Use previous month's beta as weight, start with 0 if NaN
beta_used = rolling_beta.shift(1).fillna(0.0)
df["beta_used"] = beta_used

# Beta-matched benchmark returns
df["bench_ret"] = beta_used * df["mkt_ret"] + (1.0 - beta_used) * df["rf_ret"]

df["bench_value"] = np.nan
df.loc[0, "bench_value"] = BASE_CAPITAL
for t in range(1, n):
    r_b = df.loc[t, "bench_ret"]
    df.loc[t, "bench_value"] = df.loc[t - 1, "bench_value"] * (1.0 + r_b)


# -------------------------------------------------------
# PERFORMANCE STATS
# -------------------------------------------------------
def compute_cagr(values: pd.Series, dates: pd.Series) -> float:
    v0      = values.iloc[0]
    vT      = values.iloc[-1]
    n_years = (dates.iloc[-1] - dates.iloc[0]).days / 365.25
    return (vT / v0) ** (1.0 / n_years) - 1.0

def compute_annual_vol(monthly_rets: pd.Series) -> float:
    return monthly_rets.std() * np.sqrt(12)

def compute_max_drawdown(values: pd.Series) -> float:
    running_max = values.cummax()
    dd          = values / running_max - 1.0
    return dd.min()

def compute_hist_var(monthly_rets: pd.Series, alpha: float = 0.95) -> float:
    return np.percentile(monthly_rets.dropna(), (1 - alpha) * 100)

def compute_beta(port_rets: pd.Series, bench_rets: pd.Series) -> float:
    x = bench_rets.dropna()
    y = port_rets.loc[x.index].dropna()
    common_idx = x.index.intersection(y.index)
    x = x.loc[common_idx]
    y = y.loc[common_idx]
    if len(x) < 2:
        return np.nan
    cov = np.cov(y, x)[0, 1]
    var = np.var(x)
    return cov / var if var != 0 else np.nan

rets = df.dropna(subset=["opt_ret", "bench_ret", "mkt_ret", "rf_ret"]).copy()

# Overlay
cagr_opt = compute_cagr(rets["opt_value"], rets[DATE_COL])
vol_opt  = compute_annual_vol(rets["opt_ret"])

excess_opt_month = rets["opt_ret"] - rets["rf_ret"]
mean_ex_opt      = excess_opt_month.mean()
std_opt          = rets["opt_ret"].std()

beta_opt = compute_beta(rets["opt_ret"], rets["mkt_ret"])
avg_beta = df["beta_used"].loc[rets.index].mean()

treynor_opt = np.nan
if beta_opt is not None and not np.isnan(beta_opt) and abs(beta_opt) > 1e-6 and std_opt > 0:
    treynor_opt = (mean_ex_opt * 12.0) / beta_opt

mdd_opt   = compute_max_drawdown(rets["opt_value"])
var95_opt = compute_hist_var(rets["opt_ret"], alpha=0.95)

# Beta-matched benchmark
cagr_bm = compute_cagr(rets["bench_value"], rets[DATE_COL])
vol_bm  = compute_annual_vol(rets["bench_ret"])

excess_bm_month = rets["bench_ret"] - rets["rf_ret"]
mean_ex_bm      = excess_bm_month.mean()
std_bm          = rets["bench_ret"].std()

beta_bm = compute_beta(rets["bench_ret"], rets["mkt_ret"])

treynor_bm = np.nan
if beta_bm is not None and not np.isnan(beta_bm) and abs(beta_bm) > 1e-6 and std_bm > 0:
    treynor_bm = (mean_ex_bm * 12.0) / beta_bm

mdd_bm   = compute_max_drawdown(rets["bench_value"])
var95_bm = compute_hist_var(rets["bench_ret"], alpha=0.95)

print("\n===== Performance Summary (Overlay vs Beta-Matched Benchmark) =====")
print("Overlay (RF-funded ATM calls):")
print(f"  CAGR:                         {cagr_opt*100:6.2f}%")
print(f"  Vol (ann):                    {vol_opt*100:6.2f}%")
print(f"  Treynor (rf = 1M yield):      {treynor_opt:8.4f}")
print(f"  Max Drawdown:                 {mdd_opt*100:6.2f}%")
print(f"  VaR 95% (monthly):            {var95_opt*100:6.2f}%")
print(f"  Beta vs SPX (full sample):    {beta_opt:6.3f}")
print(f"  Avg rolling beta:             {avg_beta:6.3f}")

print("\nBeta-matched Benchmark (w_t = beta_t in SPX):")
print(f"  CAGR:                         {cagr_bm*100:6.2f}%")
print(f"  Vol (ann):                    {vol_bm*100:6.2f}%")
print(f"  Treynor (rf = 1M yield):      {treynor_bm:8.4f}")
print(f"  Max Drawdown:                 {mdd_bm*100:6.2f}%")
print(f"  VaR 95% (monthly):            {var95_bm*100:6.2f}%")
print(f"  Beta vs SPX (full sample):    {beta_bm:6.3f}")

print("\nFinal overlay value:   ${:,.2f}".format(df["opt_value"].iloc[-1]))
print("Final benchmark value: ${:,.2f}".format(df["bench_value"].iloc[-1]))
print("Difference (overlay - benchmark): ${:,.2f}".format(
    df["opt_value"].iloc[-1] - df["bench_value"].iloc[-1]
))

# -------------------------------------------------------
# OPTION DIAGNOSTICS
# -------------------------------------------------------
valid_budget = df["call_budget"].dropna()
valid_cost   = df["call_cost_pct"].dropna()
valid_part   = df["opt_particip"].dropna()

total_premiums = valid_budget.sum()
total_payoffs  = df["call_payoff"].sum()
net_option_pnl = total_payoffs - total_premiums

trade_budgets = valid_budget[valid_budget > 0]
trade_parts   = valid_part[valid_part > 0]

print("\n===== Option Budget, Premiums, Participation =====")
print(f"Total premiums paid:                  ${total_premiums:,.2f}")
print(f"Total option payoffs received:        ${total_payoffs:,.2f}")
print(f"Net option P&L (payoff - premium):    ${net_option_pnl:,.2f}")
print(f"Mean monthly premium (trade months):  ${trade_budgets.mean():,.2f}")
print(f"Median monthly premium:               ${trade_budgets.median():,.2f}")
print(f"Min / Max monthly premium:            ${trade_budgets.min():,.2f} / "
      f"${trade_budgets.max():,.2f}")
print(f"Mean call cost as % of spot:          {valid_cost.mean()*100:6.2f}%")
print(f"Median call cost as % of spot:        {valid_cost.median()*100:6.2f}%")
print(f"Min / Max call cost as % of spot:     {valid_cost.min()*100:6.2f}% / "
      f"{valid_cost.max()*100:6.2f}%")
print(f"Mean participation (trade months):    {trade_parts.mean()*100:6.2f}%")
print(f"Median participation:                 {trade_parts.median()*100:6.2f}%")

print("\nSample months (first 10) with option details:")
sample_cols = [
    DATE_COL, "spx_price", "vix_price", "rf_yield_pct",
    "opt_value", "call_budget", "call_price",
    "call_cost_pct", "call_units", "call_notional",
    "opt_particip", "call_payoff", "opt_trade_pnl"
]
sample = df[sample_cols].head(10).copy()
sample["call_cost_pct"] = sample["call_cost_pct"] * 100.0
sample["opt_particip"]  = sample["opt_particip"] * 100.0
print(sample.to_string(index=False))

# Annual returns
df_year = df.set_index(DATE_COL)[["opt_value", "bench_value"]].dropna()
year_end_vals = df_year.resample("YE").last()
annual_ret = year_end_vals.pct_change().dropna()
annual_ret.index = annual_ret.index.year

print("\n===== Annual Returns (%) =====")
print((annual_ret * 100).round(2).to_string())


# -------------------------------------------------------
# PLOTS (all lines, no histograms)
# -------------------------------------------------------
def dollar_fmt(x, pos):
    return f"${x:,.0f}"

# Equity curves
plt.figure(figsize=(10, 6))
plt.plot(df[DATE_COL], df["opt_value"],   label="Overlay (RF + ATM calls)")
plt.plot(df[DATE_COL], df["bench_value"], label="Beta-matched benchmark")
ax = plt.gca()
ax.yaxis.set_major_formatter(FuncFormatter(dollar_fmt))
plt.xlabel("Date")
plt.ylabel("Portfolio Value")
plt.title("Equity Curves: Overlay vs Beta-Matched Benchmark (from Jun 2022)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Drawdowns
df["opt_running_max"]   = df["opt_value"].cummax()
df["opt_dd"]            = df["opt_value"] / df["opt_running_max"] - 1.0
df["bench_running_max"] = df["bench_value"].cummax()
df["bench_dd"]          = df["bench_value"] / df["bench_running_max"] - 1.0

plt.figure(figsize=(10, 6))
plt.plot(df[DATE_COL], df["opt_dd"],   label="Overlay drawdown")
plt.plot(df[DATE_COL], df["bench_dd"], label="Benchmark drawdown")
plt.xlabel("Date")
plt.ylabel("Drawdown")
plt.title("Drawdown Profiles (from Jun 2022)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Participation
plt.figure(figsize=(10, 4))
plt.plot(df[DATE_COL], df["opt_particip"] * 100.0)
plt.xlabel("Date")
plt.ylabel("Participation (% of portfolio)")
plt.title("Effective Call Notional / Portfolio")
plt.grid(True)
plt.tight_layout()
plt.show()

# Monthly premiums (line)
plt.figure(figsize=(10, 4))
plt.plot(df[DATE_COL], df["call_budget"].fillna(0.0))
ax = plt.gca()
ax.yaxis.set_major_formatter(FuncFormatter(dollar_fmt))
plt.xlabel("Date")
plt.ylabel("Premium paid (per month)")
plt.title("Monthly Call Premiums (funded by portfolio interest)")
plt.grid(True)
plt.tight_layout()
plt.show()

# Monthly option P&L (line, all months)
plt.figure(figsize=(10, 4))
plt.plot(df[DATE_COL], df["opt_trade_pnl"], marker="o")
ax = plt.gca()
ax.yaxis.set_major_formatter(FuncFormatter(dollar_fmt))
plt.xlabel("Date")
plt.ylabel("Option P&L (payoff - premium)")
plt.title("Monthly Option Trade P&L")
plt.grid(True)
plt.tight_layout()
plt.show()

# Rolling beta
plt.figure(figsize=(10, 4))
plt.plot(df[DATE_COL], rolling_beta, label="Rolling beta (overlay vs SPX)")
plt.xlabel("Date")
plt.ylabel("Beta")
plt.title("Rolling Beta of Overlay vs SPX")
plt.grid(True)
plt.tight_layout()
plt.show()










```
