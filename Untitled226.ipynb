```python
# =========================================================
# FULL PROJECT: NO REMOVALS, ONLY ADDITIONS
# Keeps:
# - raw file range checks
# - native Change % vs implied return cross-checks
# - weekly resample range checks
# - common sample detection
# - weekly outlier checks
# - final backtest sample diagnostics
# - semi-annual rebalancing
# - CAGR / VaR / Max DD / Calmar / Alpha / Beta
# - alpha regression summary
# - last 10 portfolio returns
# - first 5 rows of final ZAR return matrix
# - last rebalanced weights used
# - cumulative growth chart
#
# Adds:
# - 1.5% annual dividend to RSP and IWDC
# - drawdown chart
# - calendar-year return bar chart vs JSE
# - annual return histogram vs JSE
# - rolling beta
# - conditional beta (up/down)
# - downside beta
# - downside alpha (weekly and annualized)
# - information ratio
# - skewness
# - kurtosis
# - rolling alpha
# - rolling tracking error
# - professional chart styling / end-point labels
# =========================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# =========================================================
# STYLE
# =========================================================
plt.style.use("seaborn-v0_8-darkgrid")
plt.rcParams["figure.figsize"] = (13, 6)
plt.rcParams["axes.titlesize"] = 15
plt.rcParams["axes.labelsize"] = 12
plt.rcParams["legend.fontsize"] = 11
plt.rcParams["xtick.labelsize"] = 10
plt.rcParams["ytick.labelsize"] = 10

# =========================================================
# 1) FILES
# =========================================================
FILES = {
    "RSP": "RSP Historical Data (3).csv",                     # USD asset
    "QQQ_ZAR": "STXNDQJ Historical Data.csv",                 # already ZAR
    "IWDC": "IWDC Historical Data (2).csv",                   # treat via USD/ZAR per your instruction
    "TRU": "TRUJ Historical Data (5).csv",                    # already ZAR
    "BP": "BP Historical Data.csv",                           # USD asset
    "EM": "STXEMGJ Historical Data (6).csv",                  # already ZAR
    "GSBD": "GSBD Historical Data (3).csv",                   # USD asset
    "USDZAR": "USD_ZAR Historical Data (12).csv",             # FX driver
    "JSE": "FTSE_JSE All Share Historical Data (12).csv",     # benchmark
}

# =========================================================
# 2) PORTFOLIO DEFINITION
# =========================================================
TARGET_WEIGHTS = {
    "RSP": 0.20,
    "QQQ_ZAR": 0.20,
    "IWDC": 0.15,
    "TRU": 0.075,
    "BP": 0.075,
    "EM": 0.15,
    "GSBD": 0.15,
}

# Annual net dividend assumptions
ANNUAL_DIVIDENDS = {
    "RSP": 0.015,      # added
    "IWDC": 0.015,     # added
    "TRU": 0.07,
    "BP": 0.04,
    "GSBD": 0.12,
    "JSE": 0.0281,
}

REBALANCE_EVERY_WEEKS = 26
OUTLIER_THRESHOLD = 0.20
VAR_CONFIDENCE = 0.95
ROLLING_WINDOW = 52

# =========================================================
# 3) HELPERS
# =========================================================
def clean_numeric(series: pd.Series) -> pd.Series:
    return pd.to_numeric(
        series.astype(str)
              .str.replace(",", "", regex=False)
              .str.replace("%", "", regex=False)
              .str.strip(),
        errors="coerce"
    )

def load_investing_csv(path: str) -> pd.DataFrame:
    df = pd.read_csv(path)
    df.columns = [c.strip() for c in df.columns]

    if "Date" not in df.columns or "Price" not in df.columns:
        raise ValueError(f"{path} is missing required columns.")

    df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

    for col in ["Price", "Open", "High", "Low"]:
        if col in df.columns:
            df[col] = clean_numeric(df[col])

    if "Change %" in df.columns:
        df["Change_pct_file"] = clean_numeric(df["Change %"]) / 100.0
    else:
        df["Change_pct_file"] = np.nan

    df = df.dropna(subset=["Date", "Price"]).copy()
    df = df.sort_values("Date")
    df = df.drop_duplicates(subset=["Date"], keep="last")
    df = df.set_index("Date")

    return df

def resample_to_weekly_close(price_series: pd.Series) -> pd.Series:
    return price_series.resample("W-SUN").last().dropna()

def validate_native_file(df: pd.DataFrame, name: str) -> None:
    implied = df["Price"].pct_change()
    file_ret = df["Change_pct_file"]

    compare = pd.concat([implied.rename("implied"), file_ret.rename("file")], axis=1).dropna()
    if compare.empty:
        print(f"{name}: no native Change % comparison available.")
        return

    compare["abs_diff"] = (compare["implied"] - compare["file"]).abs()
    max_abs_diff = compare["abs_diff"].max()
    bad_rows = compare[compare["abs_diff"] > 0.005]

    print(f"{name}: max native return mismatch vs file Change % = {max_abs_diff:.6f}")
    if not bad_rows.empty:
        print(f"{name}: rows with mismatch > 0.5%")
        print(bad_rows.head(10))
    else:
        print(f"{name}: native return cross-check passed.")

def flag_weekly_outliers(weekly_returns: pd.Series, name: str, threshold: float = 0.20) -> None:
    flagged = weekly_returns[weekly_returns.abs() > threshold]
    if flagged.empty:
        print(f"{name}: no weekly moves above {threshold:.0%}.")
    else:
        print(f"{name}: weekly moves above {threshold:.0%}:")
        print(flagged)

def annual_to_weekly_yield(annual_yield: float) -> float:
    return annual_yield / 52.0

def cagr(returns: pd.Series) -> float:
    returns = returns.dropna()
    return (1.0 + returns).prod() ** (52.0 / len(returns)) - 1.0

def max_drawdown(returns: pd.Series) -> float:
    wealth = (1.0 + returns).cumprod()
    running_peak = wealth.cummax()
    drawdown = wealth / running_peak - 1.0
    return drawdown.min()

def historical_var(returns: pd.Series, confidence: float = 0.95) -> float:
    returns = returns.dropna()
    return -np.quantile(returns, 1.0 - confidence)

def calmar_ratio(returns: pd.Series) -> float:
    mdd = max_drawdown(returns)
    if mdd == 0:
        return np.nan
    return cagr(returns) / abs(mdd)

def jensen_alpha_annualized(portfolio_returns: pd.Series, benchmark_returns: pd.Series):
    aligned = pd.concat(
        [portfolio_returns.rename("portfolio"), benchmark_returns.rename("benchmark")],
        axis=1
    ).dropna()

    X = sm.add_constant(aligned["benchmark"].values)
    y = aligned["portfolio"].values
    model = sm.OLS(y, X).fit()

    weekly_alpha = model.params[0]
    beta = model.params[1]
    annual_alpha = weekly_alpha * 52.0

    return annual_alpha, beta, model

def information_ratio(portfolio_returns: pd.Series, benchmark_returns: pd.Series) -> float:
    aligned = pd.concat([portfolio_returns, benchmark_returns], axis=1).dropna()
    active = aligned.iloc[:, 0] - aligned.iloc[:, 1]
    te = active.std(ddof=1) * np.sqrt(52)
    if te == 0:
        return np.nan
    return (active.mean() * 52) / te

def downside_beta(portfolio_returns: pd.Series, benchmark_returns: pd.Series) -> float:
    aligned = pd.concat(
        [portfolio_returns.rename("portfolio"), benchmark_returns.rename("benchmark")],
        axis=1
    ).dropna()
    down = aligned[aligned["benchmark"] < 0].copy()
    if len(down) < 2:
        return np.nan
    bench_var = np.var(down["benchmark"].values, ddof=1)
    if bench_var == 0:
        return np.nan
    cov = np.cov(down["portfolio"].values, down["benchmark"].values, ddof=1)[0, 1]
    return cov / bench_var

def downside_alpha_annualized(portfolio_returns: pd.Series, benchmark_returns: pd.Series):
    aligned = pd.concat(
        [portfolio_returns.rename("portfolio"), benchmark_returns.rename("benchmark")],
        axis=1
    ).dropna()
    down = aligned[aligned["benchmark"] < 0].copy()
    if len(down) < 3:
        return np.nan, np.nan, None
    X = sm.add_constant(down["benchmark"].values)
    y = down["portfolio"].values
    model = sm.OLS(y, X).fit()
    weekly_alpha = model.params[0]
    annual_alpha = weekly_alpha * 52.0
    dbeta = model.params[1]
    return annual_alpha, dbeta, model

def rolling_beta(portfolio_returns: pd.Series, benchmark_returns: pd.Series, window: int = 52) -> pd.Series:
    return portfolio_returns.rolling(window).cov(benchmark_returns) / benchmark_returns.rolling(window).var()

def rolling_alpha_annualized(portfolio_returns: pd.Series, benchmark_returns: pd.Series, window: int = 52) -> pd.Series:
    aligned = pd.concat(
        [portfolio_returns.rename("portfolio"), benchmark_returns.rename("benchmark")],
        axis=1
    ).dropna()
    out = pd.Series(index=aligned.index, dtype=float)

    for i in range(window - 1, len(aligned)):
        sub = aligned.iloc[i - window + 1:i + 1]
        X = sm.add_constant(sub["benchmark"].values)
        y = sub["portfolio"].values
        model = sm.OLS(y, X).fit()
        out.iloc[i] = model.params[0] * 52.0

    return out

def annualized_tracking_error(portfolio_returns: pd.Series, benchmark_returns: pd.Series) -> float:
    aligned = pd.concat([portfolio_returns, benchmark_returns], axis=1).dropna()
    active = aligned.iloc[:, 0] - aligned.iloc[:, 1]
    return active.std(ddof=1) * np.sqrt(52)

def rolling_tracking_error(portfolio_returns: pd.Series, benchmark_returns: pd.Series, window: int = 52) -> pd.Series:
    aligned = pd.concat([portfolio_returns, benchmark_returns], axis=1).dropna()
    active = aligned.iloc[:, 0] - aligned.iloc[:, 1]
    return active.rolling(window).std(ddof=1) * np.sqrt(52)

def annotate_last(ax, x, y, fmt="{:.2f}", xoffset=6, yoffset=0):
    if len(x) == 0 or len(y) == 0:
        return
    ax.annotate(
        fmt.format(y[-1]),
        xy=(x[-1], y[-1]),
        xytext=(xoffset, yoffset),
        textcoords="offset points",
        va="center"
    )

# =========================================================
# 4) LOAD RAW DATA
# =========================================================
raw = {name: load_investing_csv(path) for name, path in FILES.items()}

print("=== RAW FILE RANGES ===")
for name, df in raw.items():
    print(f"{name}: {df.index.min().date()} to {df.index.max().date()} | rows = {len(df)}")

print("\n=== NATIVE FILE CROSS-CHECKS ===")
for name, df in raw.items():
    validate_native_file(df, name)
    print()

# =========================================================
# 5) RESAMPLE ALL PRICE SERIES TO THE SAME WEEKLY FREQUENCY
# =========================================================
weekly_prices = {}
for name, df in raw.items():
    weekly_prices[name] = resample_to_weekly_close(df["Price"])

print("=== WEEKLY RANGES AFTER RESAMPLING ===")
for name, s in weekly_prices.items():
    print(f"{name}: {s.index.min().date()} to {s.index.max().date()} | rows = {len(s)}")

# =========================================================
# 6) FIND COMMON FULL-HISTORY START DATE
# =========================================================
common_index = None
for s in weekly_prices.values():
    idx = s.dropna().index
    common_index = idx if common_index is None else common_index.intersection(idx)

common_index = common_index.sort_values()

print("\n=== COMMON SAMPLE ===")
print(f"Start date: {common_index.min().date()}")
print(f"End date  : {common_index.max().date()}")
print(f"Weeks     : {len(common_index)}")

# =========================================================
# 7) WEEKLY PRICE RETURNS
# =========================================================
weekly_price_returns = {}
for name, s in weekly_prices.items():
    weekly_price_returns[name] = s.pct_change()

print("\n=== WEEKLY OUTLIER CHECKS ON PRICE RETURNS ===")
for name, r in weekly_price_returns.items():
    flag_weekly_outliers(r.loc[common_index], name, OUTLIER_THRESHOLD)
    print()

# =========================================================
# 8) BUILD ZAR RETURNS EXACTLY AS REQUESTED
# =========================================================
usd_zar = weekly_price_returns["USDZAR"].loc[common_index]

asset_returns_zar = pd.DataFrame(index=common_index)

# USD-linked assets converted into ZAR using USD/ZAR
asset_returns_zar["RSP"] = (1.0 + weekly_price_returns["RSP"].loc[common_index]) * (1.0 + usd_zar) - 1.0
asset_returns_zar["IWDC"] = (1.0 + weekly_price_returns["IWDC"].loc[common_index]) * (1.0 + usd_zar) - 1.0
asset_returns_zar["BP"] = (1.0 + weekly_price_returns["BP"].loc[common_index]) * (1.0 + usd_zar) - 1.0
asset_returns_zar["GSBD"] = (1.0 + weekly_price_returns["GSBD"].loc[common_index]) * (1.0 + usd_zar) - 1.0

# Already ZAR-denominated assets
asset_returns_zar["QQQ_ZAR"] = weekly_price_returns["QQQ_ZAR"].loc[common_index]
asset_returns_zar["TRU"] = weekly_price_returns["TRU"].loc[common_index]
asset_returns_zar["EM"] = weekly_price_returns["EM"].loc[common_index]

# =========================================================
# 9) ADD DIVIDEND YIELDS TO TOTAL RETURNS
# =========================================================
asset_returns_zar["RSP"] = asset_returns_zar["RSP"] + annual_to_weekly_yield(ANNUAL_DIVIDENDS["RSP"])
asset_returns_zar["IWDC"] = asset_returns_zar["IWDC"] + annual_to_weekly_yield(ANNUAL_DIVIDENDS["IWDC"])
asset_returns_zar["TRU"] = asset_returns_zar["TRU"] + annual_to_weekly_yield(ANNUAL_DIVIDENDS["TRU"])
asset_returns_zar["BP"] = asset_returns_zar["BP"] + annual_to_weekly_yield(ANNUAL_DIVIDENDS["BP"])
asset_returns_zar["GSBD"] = asset_returns_zar["GSBD"] + annual_to_weekly_yield(ANNUAL_DIVIDENDS["GSBD"])

benchmark_returns = weekly_price_returns["JSE"].loc[common_index] + annual_to_weekly_yield(ANNUAL_DIVIDENDS["JSE"])

# Drop first-row NaNs from pct_change
combined = pd.concat(
    [asset_returns_zar, benchmark_returns.rename("JSE_BENCH")],
    axis=1
).dropna()

asset_returns_zar = combined[list(TARGET_WEIGHTS.keys())].copy()
benchmark_returns = combined["JSE_BENCH"].copy()

print("\n=== FINAL BACKTEST SAMPLE ===")
print(f"Portfolio sample start: {asset_returns_zar.index.min().date()}")
print(f"Portfolio sample end  : {asset_returns_zar.index.max().date()}")
print(f"Observations          : {len(asset_returns_zar)}")

# =========================================================
# 10) SEMI-ANNUAL REBALANCING BACKTEST
# =========================================================
target_weights = pd.Series(TARGET_WEIGHTS, dtype=float)

if not np.isclose(target_weights.sum(), 1.0):
    raise ValueError(f"Weights sum to {target_weights.sum():.6f}, not 1.0.")

portfolio_returns = []
weight_history = []

current_weights = target_weights.copy()

for i, dt in enumerate(asset_returns_zar.index):
    r = asset_returns_zar.loc[dt]

    port_ret = float((current_weights * r).sum())
    portfolio_returns.append(port_ret)

    wh = current_weights.copy()
    wh.name = dt
    weight_history.append(wh)

    gross_weights = current_weights * (1.0 + r)
    current_weights = gross_weights / gross_weights.sum()

    if (i + 1) % REBALANCE_EVERY_WEEKS == 0:
        current_weights = target_weights.copy()

portfolio_returns = pd.Series(portfolio_returns, index=asset_returns_zar.index, name="Portfolio")
weight_history = pd.DataFrame(weight_history)

# =========================================================
# 11) PERFORMANCE METRICS
# =========================================================
portfolio_cagr = cagr(portfolio_returns)
portfolio_var95 = historical_var(portfolio_returns, VAR_CONFIDENCE)
portfolio_mdd = max_drawdown(portfolio_returns)
portfolio_calmar = calmar_ratio(portfolio_returns)
portfolio_alpha_ann, portfolio_beta, alpha_model = jensen_alpha_annualized(portfolio_returns, benchmark_returns)

# Added metrics
portfolio_ir = information_ratio(portfolio_returns, benchmark_returns)
portfolio_skew = portfolio_returns.skew()
portfolio_kurt = portfolio_returns.kurt()
portfolio_te = annualized_tracking_error(portfolio_returns, benchmark_returns)

# Conditional betas
aligned_cond = pd.concat(
    [portfolio_returns.rename("portfolio"), benchmark_returns.rename("benchmark")],
    axis=1
).dropna()

up = aligned_cond["benchmark"] > 0
down = aligned_cond["benchmark"] < 0

beta_up = np.cov(aligned_cond.loc[up, "portfolio"], aligned_cond.loc[up, "benchmark"], ddof=1)[0, 1] / np.var(aligned_cond.loc[up, "benchmark"], ddof=1) if up.sum() > 1 else np.nan
beta_down = np.cov(aligned_cond.loc[down, "portfolio"], aligned_cond.loc[down, "benchmark"], ddof=1)[0, 1] / np.var(aligned_cond.loc[down, "benchmark"], ddof=1) if down.sum() > 1 else np.nan

# Downside beta + downside alpha
portfolio_down_beta = downside_beta(portfolio_returns, benchmark_returns)
portfolio_down_alpha_ann, portfolio_down_beta_reg, downside_alpha_model = downside_alpha_annualized(portfolio_returns, benchmark_returns)

print("\n=== RESULTS ===")
print(f"CAGR                    : {portfolio_cagr:.2%}")
print(f"Historical VaR 95%      : {portfolio_var95:.2%}")
print(f"Max Drawdown            : {portfolio_mdd:.2%}")
print(f"Calmar Ratio            : {portfolio_calmar:.4f}")
print(f"Annualized Alpha vs JSE : {portfolio_alpha_ann:.2%}")
print(f"Beta vs JSE             : {portfolio_beta:.4f}")
print(f"Information Ratio       : {portfolio_ir:.4f}")
print(f"Tracking Error (ann.)   : {portfolio_te:.2%}")
print(f"Skewness                : {portfolio_skew:.4f}")
print(f"Kurtosis                : {portfolio_kurt:.4f}")
print(f"Conditional Beta Up     : {beta_up:.4f}")
print(f"Conditional Beta Down   : {beta_down:.4f}")
print(f"Downside Beta           : {portfolio_down_beta:.4f}")
print(f"Downside Alpha (ann.)   : {portfolio_down_alpha_ann:.2%}")

# =========================================================
# 12) DIAGNOSTICS KEPT
# =========================================================
wealth_port = (1.0 + portfolio_returns).cumprod()
wealth_jse = (1.0 + benchmark_returns).cumprod()

print("\n=== LAST 10 PORTFOLIO RETURNS ===")
print(portfolio_returns.tail(10))

print("\n=== FIRST 5 ROWS OF FINAL ZAR RETURN MATRIX ===")
print(asset_returns_zar.head())

print("\n=== LAST REBALANCED WEIGHTS USED ===")
print(weight_history.tail(1).T)

print("\n=== ALPHA REGRESSION SUMMARY ===")
print(alpha_model.summary())

if downside_alpha_model is not None:
    print("\n=== DOWNSIDE ALPHA REGRESSION SUMMARY ===")
    print(downside_alpha_model.summary())

# =========================================================
# 13) ROLLING SERIES
# =========================================================
roll_beta = rolling_beta(portfolio_returns, benchmark_returns, ROLLING_WINDOW)
roll_alpha_ann = rolling_alpha_annualized(portfolio_returns, benchmark_returns, ROLLING_WINDOW)
roll_te = rolling_tracking_error(portfolio_returns, benchmark_returns, ROLLING_WINDOW)

# =========================================================
# 14) CALENDAR RETURNS
# =========================================================
calendar_port = portfolio_returns.resample("Y").apply(lambda x: (1.0 + x).prod() - 1.0)
calendar_jse = benchmark_returns.resample("Y").apply(lambda x: (1.0 + x).prod() - 1.0)

calendar_df = pd.concat(
    [calendar_port.rename("Portfolio"), calendar_jse.rename("JSE")],
    axis=1
)
calendar_df.index = calendar_df.index.year

print("\n=== CALENDAR YEAR RETURNS ===")
print(calendar_df)

# =========================================================
# 15) PLOTS - PROFESSIONAL STYLE
# =========================================================

# 15.1 Cumulative growth
fig, ax = plt.subplots()
ax.plot(wealth_port.index, wealth_port.values, linewidth=2.2, label="Portfolio")
ax.plot(wealth_jse.index, wealth_jse.values, linewidth=2.2, label="JSE Benchmark")
ax.set_title("Portfolio vs JSE Benchmark - Cumulative Growth of 1")
ax.set_xlabel("Date")
ax.set_ylabel("Growth of 1")
annotate_last(ax, wealth_port.index, wealth_port.values, fmt="{:.2f}")
annotate_last(ax, wealth_jse.index, wealth_jse.values, fmt="{:.2f}", yoffset=-12)
ax.legend()
plt.tight_layout()
plt.show()

# 15.2 Drawdown curve
drawdown_port = wealth_port / wealth_port.cummax() - 1.0
drawdown_jse = wealth_jse / wealth_jse.cummax() - 1.0

fig, ax = plt.subplots()
ax.plot(drawdown_port.index, drawdown_port.values, linewidth=2.2, label="Portfolio")
ax.plot(drawdown_jse.index, drawdown_jse.values, linewidth=2.0, label="JSE Benchmark")
ax.set_title("Drawdown Curve")
ax.set_xlabel("Date")
ax.set_ylabel("Drawdown")
annotate_last(ax, drawdown_port.index, drawdown_port.values, fmt="{:.2%}")
annotate_last(ax, drawdown_jse.index, drawdown_jse.values, fmt="{:.2%}", yoffset=-12)
ax.legend()
plt.tight_layout()
plt.show()

# 15.3 Calendar year return bars
fig, ax = plt.subplots(figsize=(14, 6))
x = np.arange(len(calendar_df.index))
width = 0.38

ax.bar(x - width / 2, calendar_df["Portfolio"].values, width=width, label="Portfolio")
ax.bar(x + width / 2, calendar_df["JSE"].values, width=width, label="JSE")
ax.axhline(0, linewidth=1)
ax.set_xticks(x)
ax.set_xticklabels(calendar_df.index.astype(str), rotation=0)
ax.set_title("Calendar Year Returns - Portfolio vs JSE")
ax.set_xlabel("Year")
ax.set_ylabel("Return")
ax.legend()
plt.tight_layout()
plt.show()

# 15.4 Annual return distribution histogram
fig, ax = plt.subplots()
ax.hist(calendar_df["Portfolio"].dropna().values, alpha=0.65, label="Portfolio")
ax.hist(calendar_df["JSE"].dropna().values, alpha=0.65, label="JSE")
ax.set_title("Annual Return Distribution")
ax.set_xlabel("Annual Return")
ax.set_ylabel("Frequency")
ax.legend()
plt.tight_layout()
plt.show()

# 15.5 Rolling beta
fig, ax = plt.subplots()
ax.plot(roll_beta.index, roll_beta.values, linewidth=2.2)
ax.axhline(0, linewidth=1)
ax.set_title(f"Rolling Beta vs JSE ({ROLLING_WINDOW}-Week)")
ax.set_xlabel("Date")
ax.set_ylabel("Beta")
roll_beta_clean = roll_beta.dropna()
if len(roll_beta_clean) > 0:
    annotate_last(ax, roll_beta_clean.index, roll_beta_clean.values, fmt="{:.2f}")
plt.tight_layout()
plt.show()

# 15.6 Rolling alpha annualized
fig, ax = plt.subplots()
ax.plot(roll_alpha_ann.index, roll_alpha_ann.values, linewidth=2.2)
ax.axhline(0, linewidth=1)
ax.set_title(f"Rolling Alpha vs JSE - Annualized ({ROLLING_WINDOW}-Week)")
ax.set_xlabel("Date")
ax.set_ylabel("Annualized Alpha")
roll_alpha_clean = roll_alpha_ann.dropna()
if len(roll_alpha_clean) > 0:
    annotate_last(ax, roll_alpha_clean.index, roll_alpha_clean.values, fmt="{:.2%}")
plt.tight_layout()
plt.show()

# 15.7 Rolling tracking error
fig, ax = plt.subplots()
ax.plot(roll_te.index, roll_te.values, linewidth=2.2)
ax.set_title(f"Rolling Tracking Error - Annualized ({ROLLING_WINDOW}-Week)")
ax.set_xlabel("Date")
ax.set_ylabel("Tracking Error")
roll_te_clean = roll_te.dropna()
if len(roll_te_clean) > 0:
    annotate_last(ax, roll_te_clean.index, roll_te_clean.values, fmt="{:.2%}")
plt.tight_layout()
plt.show()

# 15.8 Scatter with regression line
aligned_plot = pd.concat(
    [portfolio_returns.rename("Portfolio"), benchmark_returns.rename("JSE")],
    axis=1
).dropna()

coef = np.polyfit(aligned_plot["JSE"].values, aligned_plot["Portfolio"].values, 1)
xline = np.linspace(aligned_plot["JSE"].min(), aligned_plot["JSE"].max(), 100)
yline = coef[0] * xline + coef[1]

fig, ax = plt.subplots()
ax.scatter(aligned_plot["JSE"].values, aligned_plot["Portfolio"].values, alpha=0.5)
ax.plot(xline, yline, linewidth=2.2)
ax.axhline(0, linewidth=1)
ax.axvline(0, linewidth=1)
ax.set_title("Portfolio vs JSE Weekly Returns")
ax.set_xlabel("JSE Weekly Return")
ax.set_ylabel("Portfolio Weekly Return")
plt.tight_layout()
plt.show()

# =========================================================
# 16) CONDITIONAL / DOWNSIDE SUMMARY TABLE
# =========================================================
summary_table = pd.DataFrame({
    "Metric": [
        "CAGR",
        "VaR 95%",
        "Max Drawdown",
        "Calmar",
        "Alpha vs JSE (ann.)",
        "Beta vs JSE",
        "Information Ratio",
        "Tracking Error (ann.)",
        "Skewness",
        "Kurtosis",
        "Conditional Beta Up",
        "Conditional Beta Down",
        "Downside Beta",
        "Downside Alpha (ann.)"
    ],
    "Value": [
        portfolio_cagr,
        portfolio_var95,
        portfolio_mdd,
        portfolio_calmar,
        portfolio_alpha_ann,
        portfolio_beta,
        portfolio_ir,
        portfolio_te,
        portfolio_skew,
        portfolio_kurt,
        beta_up,
        beta_down,
        portfolio_down_beta,
        portfolio_down_alpha_ann
    ]
})

print("\n=== SUMMARY TABLE ===")
print(summary_table)

# =========================================================
# 17) WEIGHT HISTORY SNAPSHOT
# =========================================================
print("\n=== WEIGHT HISTORY TAIL ===")
print(weight_history.tail())

# Optional weight plot
fig, ax = plt.subplots(figsize=(14, 7))
for col in weight_history.columns:
    ax.plot(weight_history.index, weight_history[col].values, linewidth=1.8, label=col)
ax.set_title("Portfolio Weights Through Time")
ax.set_xlabel("Date")
ax.set_ylabel("Weight")
ax.legend(ncol=2)
plt.tight_layout()
plt.show()

# =========================================================
# 18) ALPHA OPTIMISER (PURE VERSION 1)
# =========================================================
from scipy.optimize import minimize

# Ensure alignment
aligned = pd.concat(
    [asset_returns_zar, benchmark_returns.rename("JSE")],
    axis=1
).dropna()

R = aligned[asset_returns_zar.columns]   # asset return matrix
B = aligned["JSE"]                      # benchmark

assets = R.columns
n = len(assets)

# =========================================================
# Objective: maximize annualized Jensen alpha
# =========================================================
def alpha_objective(weights):
    w = np.array(weights)

    port = (R.values @ w)

    X = sm.add_constant(B.values)
    model = sm.OLS(port, X).fit()

    weekly_alpha = model.params[0]
    annual_alpha = weekly_alpha * 52

    return -annual_alpha  # minimize negative alpha

# =========================================================
# Constraints
# =========================================================
constraints = [
    {"type": "eq", "fun": lambda w: np.sum(w) - 1.0}  # sum to 1
]

# Bounds (long-only, no leverage)
bounds = [(0.0, 1.0) for _ in range(n)]

# Initial guess = current weights
w0 = np.array(list(TARGET_WEIGHTS.values()))

# =========================================================
# Optimization
# =========================================================
result = minimize(
    alpha_objective,
    w0,
    method="SLSQP",
    bounds=bounds,
    constraints=constraints,
    options={"maxiter": 1000}
)

opt_weights = pd.Series(result.x, index=assets)

# =========================================================
# Compute optimized portfolio performance
# =========================================================
opt_port = pd.Series(R.values @ opt_weights.values, index=R.index)

opt_alpha, opt_beta, _ = jensen_alpha_annualized(opt_port, B)

# =========================================================
# Output
# =========================================================
print("\n=== OPTIMIZED WEIGHTS (ALPHA MAX) ===")
print(opt_weights.round(4))

print("\n=== OPTIMIZED PERFORMANCE ===")
print("Alpha:", round(opt_alpha,4))
print("Beta:", round(opt_beta,4))
print("CAGR:", round(cagr(opt_port),4))
print("MDD:", round(max_drawdown(opt_port),4))

# =========================================================
# Compare cumulative
# =========================================================
plt.figure(figsize=(12,6))
plt.plot((1+portfolio_returns).cumprod(), label="Original", linewidth=2)
plt.plot((1+opt_port).cumprod(), label="Optimized", linewidth=2)
plt.title("Original vs Alpha-Optimized Portfolio")
plt.legend()
plt.show()

# =========================================================
# 18) ALPHA OPTIMISER (PURE VERSION 1)
# =========================================================
from scipy.optimize import minimize

# Ensure alignment
aligned = pd.concat(
    [asset_returns_zar, benchmark_returns.rename("JSE")],
    axis=1
).dropna()

R = aligned[asset_returns_zar.columns]   # asset return matrix
B = aligned["JSE"]                      # benchmark

assets = R.columns
n = len(assets)

# =========================================================
# Objective: maximize annualized Jensen alpha
# =========================================================
def alpha_objective(weights):
    w = np.array(weights)

    port = (R.values @ w)

    X = sm.add_constant(B.values)
    model = sm.OLS(port, X).fit()

    weekly_alpha = model.params[0]
    annual_alpha = weekly_alpha * 52

    return -annual_alpha  # minimize negative alpha

# =========================================================
# Constraints
# =========================================================
constraints = [
    {"type": "eq", "fun": lambda w: np.sum(w) - 1.0}  # sum to 1
]

# Bounds (long-only, no leverage)
bounds = [(0.0, 1.0) for _ in range(n)]

# Initial guess = current weights
w0 = np.array(list(TARGET_WEIGHTS.values()))

# =========================================================
# Optimization
# =========================================================
result = minimize(
    alpha_objective,
    w0,
    method="SLSQP",
    bounds=bounds,
    constraints=constraints,
    options={"maxiter": 1000}
)

opt_weights = pd.Series(result.x, index=assets)

# =========================================================
# Compute optimized portfolio performance
# =========================================================
opt_port = pd.Series(R.values @ opt_weights.values, index=R.index)

opt_alpha, opt_beta, _ = jensen_alpha_annualized(opt_port, B)

# =========================================================
# Output
# =========================================================
print("\n=== OPTIMIZED WEIGHTS (ALPHA MAX) ===")
print(opt_weights.round(4))

print("\n=== OPTIMIZED PERFORMANCE ===")
print("Alpha:", round(opt_alpha,4))
print("Beta:", round(opt_beta,4))
print("CAGR:", round(cagr(opt_port),4))
print("MDD:", round(max_drawdown(opt_port),4))

# =========================================================
# Compare cumulative
# =========================================================
plt.figure(figsize=(12,6))
plt.plot((1+portfolio_returns).cumprod(), label="Original", linewidth=2)
plt.plot((1+opt_port).cumprod(), label="Optimized", linewidth=2)
plt.title("Original vs Alpha-Optimized Portfolio")
plt.legend()
plt.show()
```
