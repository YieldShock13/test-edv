```python
#  Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy.stats import norm, spearmanr, kendalltau
from statsmodels.distributions.empirical_distribution import ECDF

#  Load Data (assuming files are uploaded to Colab)
btc = pd.read_csv('/content/Bitcoin Historical Data.csv')
spy = pd.read_csv('/content/S&P 500 Historical Data (2).csv')
gold = pd.read_csv('/content/XAU_USD Historical Data (1).csv')
dxy = pd.read_csv('/content/US Dollar Index Historical Data (1).csv')

#  Clean & Merge
for df in [btc, spy, gold, dxy]:
    df.columns = [col.strip() for col in df.columns]
    df['Date'] = pd.to_datetime(df['Date'])
    df['Price'] = pd.to_numeric(df['Price'].astype(str).str.replace(',', ''), errors='coerce')
    df.dropna(subset=['Date', 'Price'], inplace=True)

btc = btc[['Date', 'Price']].rename(columns={'Price': 'BTC'})
spy = spy[['Date', 'Price']].rename(columns={'Price': 'SPY'})
gold = gold[['Date', 'Price']].rename(columns={'Price': 'Gold'})
dxy = dxy[['Date', 'Price']].rename(columns={'Price': 'DXY'})

df = btc.merge(spy, on='Date').merge(gold, on='Date').merge(dxy, on='Date').set_index('Date')
df = df.sort_index()  # Ensure chronological order for rolling operations
df = df.dropna()

#  Log Returns
returns = np.log(df / df.shift(1)).dropna()

#  1. Rolling Spearman Correlation (12-month) + Pseudo R
window = 12
plt.figure(figsize=(14, 8))
for a, b in [('BTC', 'SPY'), ('BTC', 'Gold'), ('BTC', 'DXY')]:
    # Rolling Spearman: apply spearmanr on paired windows
    rs = returns[a].rolling(window).apply(lambda x: spearmanr(x, returns[b].loc[x.index])[0] if len(x) >= 2 else np.nan)
    pseudo_r2_val = (rs**2).mean()  # Pseudo R as rho squared
    plt.plot(rs, label=f'{a}-{b} Spearman Corr (avg Pseudo R = {pseudo_r2_val:.2f})')
plt.title('Rolling Spearman Correlation (12-Month Window)')
plt.xlabel('Date'); plt.ylabel('Spearman Correlation')
plt.legend(); plt.grid(); plt.tight_layout(); plt.show()

#  2. Standardized Rolling Correlation
std_returns = (returns - returns.rolling(window).mean()) / returns.rolling(window).std()
plt.figure(figsize=(14, 8))
for a, b in [('BTC', 'SPY'), ('BTC', 'Gold'), ('BTC', 'DXY')]:
    src = std_returns[a].rolling(window).corr(std_returns[b])
    r2_val = (src**2).mean()
    plt.plot(src, label=f'{a}-{b} Std Corr (avg R = {r2_val:.2f})')
plt.title('Standardized Rolling Correlation (12-Month Window)')
plt.xlabel('Date'); plt.ylabel('Std Corr')
plt.legend(); plt.grid(); plt.tight_layout(); plt.show()

#  3. Rolling Beta + R (as Histogram for Neater View)
def rolling_beta_r2(y, x, window):
    beta_all, beta_down, r2_all, r2_down = [], [], [], []
    for i in range(window, len(y)):
        xi = x[i-window:i]
        yi = y[i-window:i]
        reg = LinearRegression().fit(xi.values.reshape(-1,1), yi.values)
        beta_all.append(reg.coef_[0])
        r2_all.append(reg.score(xi.values.reshape(-1,1), yi.values))
        mask = xi < 0
        if mask.sum() >= 3:
            reg_d = LinearRegression().fit(xi[mask].values.reshape(-1,1), yi[mask].values)
            beta_down.append(reg_d.coef_[0])
            r2_down.append(reg_d.score(xi[mask].values.reshape(-1,1), yi[mask].values))
        else:
            beta_down.append(np.nan)
            r2_down.append(np.nan)
    return beta_all, beta_down, r2_all, r2_down

b_all, b_down, r2_all, r2_down = rolling_beta_r2(returns['BTC'], returns['SPY'], window)
# Histogram for betas (neater visualization)
fig, axs = plt.subplots(1, 2, figsize=(14, 6))
axs[0].hist(b_all, bins=30, alpha=0.7, color='blue', label=f'All Beta (mean: {np.nanmean(b_all):.2f}, std: {np.nanstd(b_all):.2f})')
axs[0].legend(); axs[0].set_title('Histogram of Rolling Beta (All)')
axs[0].set_xlabel('Beta Value'); axs[0].set_ylabel('Frequency')
axs[1].hist([v for v in b_down if not np.isnan(v)], bins=30, alpha=0.7, color='red', label=f'Down Beta (mean: {np.nanmean(b_down):.2f}, std: {np.nanstd(b_down):.2f})')
axs[1].legend(); axs[1].set_title('Histogram of Rolling Beta (SPY < 0)')
axs[1].set_xlabel('Beta Value'); axs[1].set_ylabel('Frequency')
plt.tight_layout(); plt.show()

#  4. Gaussian Copula (explicit transform) with Summary Stats and Kendall Tau for all pairs
# Pairs to process
pairs = [('BTC', 'SPY'), ('BTC', 'Gold'), ('BTC', 'DXY')]

# Plot for each pair
for a, b in pairs:
    # Step 1: Empirical CDF (uniform transform)
    a_u = ECDF(returns[a])(returns[a])
    b_u = ECDF(returns[b])(returns[b])

    # Step 2: Inverse Normal CDF (to standard normal)
    a_n = norm.ppf(np.clip(a_u, 1e-6, 1 - 1e-6))
    b_n = norm.ppf(np.clip(b_u, 1e-6, 1 - 1e-6))

    # Step 3: Rolling correlation on transformed normals
    a_n_series = pd.Series(a_n, index=returns.index)
    b_n_series = pd.Series(b_n, index=returns.index)
    copula_corr = a_n_series.rolling(window).corr(b_n_series)
    pseudo_r2_val = (copula_corr**2).mean()  # Pseudo R as corr squared

    # Add Rolling Kendall Tau for Copula
    copula_kendall = returns[a].rolling(window).apply(lambda x: kendalltau(x, returns[b].loc[x.index])[0] if len(x) >= 2 else np.nan)

    # Plot Copula Corr and Kendall Tau
    plt.figure(figsize=(14, 8))
    plt.plot(copula_corr, label=f'Gaussian Copula Corr (mean: {copula_corr.mean():.2f}, std: {copula_corr.std():.2f}, avg Pseudo R = {pseudo_r2_val:.2f})', color='purple')
    plt.plot(copula_kendall, label=f'Kendall Tau (mean: {copula_kendall.mean():.2f}, std: {copula_kendall.std():.2f})', color='green', linestyle='--')
    plt.title(f'Rolling Gaussian Copula Correlation and Kendall Tau ({a} vs {b})')
    plt.xlabel('Date'); plt.ylabel('Value')
    plt.grid(); plt.legend(); plt.tight_layout(); plt.show()

    #  Preview transformations (for debugging)
    print(f"{a} ECDF (first 5):", a_u[:5])
    print(f"{b} ECDF (first 5):", b_u[:5])
    print(f"{a} to N(0,1) (first 5):", a_n[:5])
    print(f"{b} to N(0,1) (first 5):", b_n[:5])
    print(f"Copula Corr (last 5):", copula_corr.tail(5))
    print(f"Kendall Tau (last 5):", copula_kendall.tail(5))


```
