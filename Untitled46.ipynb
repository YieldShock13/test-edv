```python
# 📦 Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy.stats import norm, spearmanr, kendalltau
from statsmodels.distributions.empirical_distribution import ECDF

# Set dark background style and font (match closer to image: prioritize Arial-like sans-serif)
plt.style.use('dark_background')
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial', 'DejaVu Sans', 'Helvetica']  # Prioritize Arial/Helvetica
plt.rcParams['font.size'] = 10  # Adjust size for readability
plt.rcParams['text.color'] = 'lightblue'  # Match cyan-ish text
plt.rcParams['axes.labelcolor'] = 'lightblue'
plt.rcParams['xtick.color'] = 'lightblue'
plt.rcParams['ytick.color'] = 'lightblue'
plt.rcParams['grid.color'] = 'gray'

# 📁 Load Data (assuming files are uploaded to Colab)
btc = pd.read_csv('/content/Bitcoin Historical Data.csv')
spy = pd.read_csv('/content/S&P 500 Historical Data (2).csv')
gold = pd.read_csv('/content/XAU_USD Historical Data (1).csv')
dxy = pd.read_csv('/content/US Dollar Index Historical Data (1).csv')

# 🧼 Clean & Merge
for df in [btc, spy, gold, dxy]:
    df.columns = [col.strip() for col in df.columns]
    df['Date'] = pd.to_datetime(df['Date'])
    df['Price'] = pd.to_numeric(df['Price'].astype(str).str.replace(',', ''), errors='coerce')
    df.dropna(subset=['Date', 'Price'], inplace=True)

btc = btc[['Date', 'Price']].rename(columns={'Price': 'BTC'})
spy = spy[['Date', 'Price']].rename(columns={'Price': 'SPY'})
gold = gold[['Date', 'Price']].rename(columns={'Price': 'Gold'})
dxy = dxy[['Date', 'Price']].rename(columns={'Price': 'DXY'})

df = btc.merge(spy, on='Date').merge(gold, on='Date').merge(dxy, on='Date').set_index('Date')
df = df.sort_index()  # Ensure chronological order for rolling operations
df = df.dropna()

# 📉 Log Returns
returns = np.log(df / df.shift(1)).dropna()

# 🔁 1. Rolling Spearman Correlation (12-month) + Pseudo R²
window = 12
fig, ax = plt.subplots(figsize=(14, 8))
for a, b in [('BTC', 'SPY'), ('BTC', 'Gold'), ('BTC', 'DXY')]:
    # Rolling Spearman: apply spearmanr on paired windows
    rs = returns[a].rolling(window).apply(lambda x: spearmanr(x, returns[b].loc[x.index])[0] if len(x) >= 2 else np.nan)
    pseudo_r2_val = (rs**2).mean()  # Pseudo R² as rho squared
    ax.plot(rs, label=f'{a}-{b} Spearman Corr (avg Pseudo R² = {pseudo_r2_val:.2f})', color='cyan')
ax.set_title('Rolling Spearman Correlation (12-Month Window)')
ax.set_xlabel('Date'); ax.set_ylabel('Spearman Correlation')
ax.legend(); ax.grid();
# Add watermark
ax.text(0.95, 0.01, 'Pearl Gray Equity and Research', fontsize=10, color='gray', ha='right', va='bottom', alpha=0.5)
plt.tight_layout(); plt.show()

# 🔁 2. Standardized Rolling Correlation
std_returns = (returns - returns.rolling(window).mean()) / returns.rolling(window).std()
fig, ax = plt.subplots(figsize=(14, 8))
for a, b in [('BTC', 'SPY'), ('BTC', 'Gold'), ('BTC', 'DXY')]:
    src = std_returns[a].rolling(window).corr(std_returns[b])
    r2_val = (src**2).mean()
    ax.plot(src, label=f'{a}-{b} Std Corr (avg R² = {r2_val:.2f})', color='cyan')
ax.set_title('Standardized Rolling Correlation (12-Month Window)')
ax.set_xlabel('Date'); ax.set_ylabel('Std Corr')
ax.legend(); ax.grid();
# Add watermark
ax.text(0.95, 0.01, 'Pearl Gray Equity and Research', fontsize=10, color='gray', ha='right', va='bottom', alpha=0.5)
plt.tight_layout(); plt.show()

# 📐 3. Rolling Beta + R² (as Histogram for Neater View)
def rolling_beta_r2(y, x, window):
    beta_all, beta_down, r2_all, r2_down = [], [], [], []
    for i in range(window, len(y)):
        xi = x[i-window:i]
        yi = y[i-window:i]
        reg = LinearRegression().fit(xi.values.reshape(-1,1), yi.values)
        beta_all.append(reg.coef_[0])
        r2_all.append(reg.score(xi.values.reshape(-1,1), yi.values))
        mask = xi < 0
        if mask.sum() >= 3:
            reg_d = LinearRegression().fit(xi[mask].values.reshape(-1,1), yi[mask].values)
            beta_down.append(reg_d.coef_[0])
            r2_down.append(reg_d.score(xi[mask].values.reshape(-1,1), yi[mask].values))
        else:
            beta_down.append(np.nan)
            r2_down.append(np.nan)
    return beta_all, beta_down, r2_all, r2_down

b_all, b_down, r2_all, r2_down = rolling_beta_r2(returns['BTC'], returns['SPY'], window)
# Histogram for betas (neater visualization)
fig, axs = plt.subplots(1, 2, figsize=(14, 6))
axs[0].hist(b_all, bins=30, alpha=0.7, color='cyan', label=f'All Beta (mean: {np.nanmean(b_all):.2f}, std: {np.nanstd(b_all):.2f})')
axs[0].legend(); axs[0].set_title('Histogram of Rolling Beta (All)')
axs[0].set_xlabel('Beta Value'); axs[0].set_ylabel('Frequency')
axs[1].hist([v for v in b_down if not np.isnan(v)], bins=30, alpha=0.7, color='cyan', label=f'Down Beta (mean: {np.nanmean(b_down):.2f}, std: {np.nanstd(b_down):.2f})')
axs[1].legend(); axs[1].set_title('Histogram of Rolling Beta (SPY < 0)')
axs[1].set_xlabel('Beta Value'); axs[1].set_ylabel('Frequency')
# Add watermark to fig
fig.text(0.95, 0.01, 'Pearl Gray Equity and Research', fontsize=10, color='gray', ha='right', va='bottom', alpha=0.5)
plt.tight_layout(); plt.show()

# 📈 4. Gaussian Copula (explicit transform) with Summary Stats and Kendall Tau for all pairs
# Pairs to process
pairs = [('BTC', 'SPY'), ('BTC', 'Gold'), ('BTC', 'DXY')]

# Plot for each pair
for a, b in pairs:
    # Step 1: Empirical CDF (uniform transform)
    a_u = ECDF(returns[a])(returns[a])
    b_u = ECDF(returns[b])(returns[b])

    # Step 2: Inverse Normal CDF (to standard normal)
    a_n = norm.ppf(np.clip(a_u, 1e-6, 1 - 1e-6))
    b_n = norm.ppf(np.clip(b_u, 1e-6, 1 - 1e-6))

    # Step 3: Rolling correlation on transformed normals
    a_n_series = pd.Series(a_n, index=returns.index)
    b_n_series = pd.Series(b_n, index=returns.index)
    copula_corr = a_n_series.rolling(window).corr(b_n_series)
    pseudo_r2_val = (copula_corr**2).mean()  # Pseudo R² as corr squared

    # Add Rolling Kendall Tau for Copula
    copula_kendall = returns[a].rolling(window).apply(lambda x: kendalltau(x, returns[b].loc[x.index])[0] if len(x) >= 2 else np.nan)

    # Plot Copula Corr and Kendall Tau
    fig, ax = plt.subplots(figsize=(14, 8))
    ax.plot(copula_corr, label=f'Gaussian Copula Corr (mean: {copula_corr.mean():.2f}, std: {copula_corr.std():.2f}, avg Pseudo R² = {pseudo_r2_val:.2f})', color='cyan')
    ax.plot(copula_kendall, label=f'Kendall Tau (mean: {copula_kendall.mean():.2f}, std: {copula_kendall.std():.2f})', color='lightblue', linestyle='--')
    ax.set_title(f'Rolling Gaussian Copula Correlation and Kendall Tau ({a} vs {b})')
    ax.set_xlabel('Date'); ax.set_ylabel('Value')
    ax.grid(); ax.legend();
    # Add watermark
    ax.text(0.95, 0.01, 'Pearl Gray Equity and Research', fontsize=10, color='gray', ha='right', va='bottom', alpha=0.5)
    plt.tight_layout(); plt.show()

    # ✅ Preview transformations (for debugging)
    print(f"{a} ECDF (first 5):", a_u[:5])
    print(f"{b} ECDF (first 5):", b_u[:5])
    print(f"{a} to N(0,1) (first 5):", a_n[:5])
    print(f"{b} to N(0,1) (first 5):", b_n[:5])
    print(f"Copula Corr (last 5):", copula_corr.tail(5))
    print(f"Kendall Tau (last 5):", copula_kendall.tail(5))

# 📊 Year-by-Year Averages Table (for BTC-SPY as example; extend if needed)
# Compute metrics for BTC-SPY
pearson_corr = returns['BTC'].rolling(window).corr(returns['SPY'])
spearman_corr = returns['BTC'].rolling(window).apply(lambda x: spearmanr(x, returns['SPY'].loc[x.index])[0] if len(x) >= 2 else np.nan)
kendall_tau = returns['BTC'].rolling(window).apply(lambda x: kendalltau(x, returns['SPY'].loc[x.index])[0] if len(x) >= 2 else np.nan)
beta_all = pd.Series(b_all, index=returns.index[window:])
beta_down = pd.Series(b_down, index=returns.index[window:])

# Create yearly means
metrics_df = pd.DataFrame({
    'Pearson Corr': pearson_corr,
    'Spearman Corr': spearman_corr,
    'Kendall Tau': kendall_tau,
    'Beta All': beta_all,
    'Beta Down': beta_down
})
metrics_df['Year'] = metrics_df.index.year
yearly_avg = metrics_df.groupby('Year').mean().reset_index()
yearly_avg['Year'] = yearly_avg['Year'].astype(int)  # No decimals for years
yearly_avg.iloc[:, 1:] = yearly_avg.iloc[:, 1:].round(2)  # 2 decimals for metrics

# Display table with nice formatting and darker blue text
fig, ax = plt.subplots(figsize=(10, 6))
ax.axis('tight')
ax.axis('off')
table = ax.table(cellText=yearly_avg.values, colLabels=yearly_avg.columns, cellLoc='center', loc='center')
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2)
# Set darker blue for stats text (excluding headers)
for (i, j), cell in table.get_celld().items():
    if i == 0:  # Header row
        cell.set_text_props(color='lightblue')  # Keep header light
    else:
        cell.set_text_props(color='darkblue')  # Darker blue for stats
# Add watermark to fig
fig.text(0.95, 0.01, 'Pearl Gray Equity and Research', fontsize=10, color='gray', ha='right', va='bottom', alpha=0.5)
plt.title('Year-by-Year Averages for BTC vs SPY Metrics')
plt.show()


```
