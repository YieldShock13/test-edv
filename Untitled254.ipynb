```python
import numpy as np
import scipy as sp
import pandas as pd
import math as mh

df = pd.read_csv("SPY Historical Data.csv")
print (df)

price = df["Price"]
returns = (price/price.shift(1)).dropna()

Variance = np.var(returns)
Covariance = np.cov(returns, returns)[0, 1]

beta = Covariance/Variance

print(beta)







```
