```python
# ============================================================
# COPPER vs IRON ORE  +  MSCI CHINA vs MSCI INDIA
# Rolling 12M Z-STAT + VOL FILTER (mean-reversion sustainability)
# Monthly data | Jupyter/Colab kernel | uses filenames as-is
# Bloomberg-terminal-ish charts + watermark: "Orca's Capital Strategies"
# ============================================================

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from dateutil import parser

# -----------------------------
# FILENAMES (AS YOU PROVIDED)
# -----------------------------
F_COPPER = "PCOPPUSDM.csv"
F_IRON   = "PIORECRUSDM.csv"
F_CHINA  = "MSCI China Historical Data.csv"
F_INDIA  = "MSCI India Historical Data.csv"

# -----------------------------
# PARAMS
# -----------------------------
ROLL_Z = 12           # rolling z-stat window (months)
ROLL_VOL = 12         # rolling vol window (months)
BASELINE_VOL = 60     # baseline window for vol regime (months)
Z_CAP = 4.0           # chart cap for readability
WATERMARK = "Orca's Capital Strategies"

# ============================================================
# FILE RESOLVER (kernel-friendly)
# ============================================================
def resolve_file(fname: str) -> str:
    """
    Finds the file in common kernel locations without hardcoding /mnt/data/ etc.
    Uses the filename exactly as given.
    """
    candidates = [
        fname,
        os.path.join(".", fname),
        os.path.join("content", fname),
        os.path.join("/content", fname),
        os.path.join("data", fname),
    ]
    for p in candidates:
        if os.path.exists(p):
            return p
    # last resort: search current tree (can be slow but fine for small dirs)
    for root, _, files in os.walk("."):
        if fname in files:
            return os.path.join(root, fname)
    raise FileNotFoundError(f"Could not find file: {fname} (ensure it's uploaded into the kernel workspace)")

# ============================================================
# DATE NORMALIZATION (FIXES YOUR 'MS' PERIOD ERROR)
# ============================================================
def month_start(dt_series: pd.Series) -> pd.Series:
    """
    Convert any datetime series to month-start timestamps safely.
    NOTE: Do NOT use to_timestamp('MS') -> that breaks in current pandas.
    """
    dt = pd.to_datetime(dt_series, errors="coerce")
    # period -> timestamp start (month start)
    return dt.dt.to_period("M").dt.to_timestamp(how="start")

# ============================================================
# LOADERS
# ============================================================
def read_fred_monthly(filename: str) -> pd.DataFrame:
    """
    FRED CSV format: observation_date + 1 value column
    """
    path = resolve_file(filename)
    df = pd.read_csv(path)

    if "observation_date" not in df.columns:
        raise ValueError(f"{filename}: expected column 'observation_date'. Found: {df.columns.tolist()}")

    val_cols = [c for c in df.columns if c != "observation_date"]
    if len(val_cols) != 1:
        raise ValueError(f"{filename}: expected exactly 1 value column besides observation_date. Found: {val_cols}")

    v = val_cols[0]
    df["Date"] = month_start(df["observation_date"])
    df[v] = pd.to_numeric(df[v], errors="coerce")

    df = df.dropna(subset=["Date", v]).sort_values("Date").set_index("Date")[[v]]
    df.columns = ["Price"]
    return df

def read_investing_monthly(filename: str) -> pd.DataFrame:
    """
    Investing.com CSV typical: Date, Price, Open, High, Low, Vol., Change %
    """
    path = resolve_file(filename)
    df = pd.read_csv(path)

    if "Date" not in df.columns or "Price" not in df.columns:
        raise ValueError(f"{filename}: expected columns ['Date','Price']. Found: {df.columns.tolist()}")

    # robust date parse
    def parse_any(x):
        try:
            return parser.parse(str(x))
        except Exception:
            return pd.NaT

    df["Date"] = df["Date"].apply(parse_any)
    df["Date"] = month_start(df["Date"])

    # handle comma decimals / commas in thousands
    px = (
        df["Price"]
        .astype(str)
        .str.replace(",", "", regex=False)
        .str.strip()
    )
    df["Price"] = pd.to_numeric(px, errors="coerce")

    df = df.dropna(subset=["Date", "Price"]).sort_values("Date").set_index("Date")[["Price"]]
    return df

# ============================================================
# STATS: Z + VOL FILTER
# ============================================================
def rolling_zscore(x: pd.Series, window: int) -> pd.Series:
    mu = x.rolling(window).mean()
    sd = x.rolling(window).std(ddof=0)
    z = (x - mu) / sd
    return z.replace([np.inf, -np.inf], np.nan)

def vol_filter(x: pd.Series, vol_window: int, baseline_window: int) -> pd.DataFrame:
    """
    Volatility regime:
      VOL = rolling std of x (12M)
      BASE = rolling median of VOL (60M)
      LOW_VOL = VOL <= BASE  (better mean-reversion regime)
    """
    vol = x.rolling(vol_window).std(ddof=0)
    base = vol.rolling(baseline_window).median()
    return pd.DataFrame({"VOL": vol, "VOL_BASE": base, "LOW_VOL": (vol <= base)})

# ============================================================
# BLOOMBERG-ISH STYLE + WATERMARK
# ============================================================
def set_bbg_style():
    plt.rcParams.update({
        "figure.facecolor": "#0b0f14",
        "axes.facecolor":   "#0b0f14",
        "axes.edgecolor":   "#6b7280",
        "axes.labelcolor":  "#e5e7eb",
        "text.color":       "#e5e7eb",
        "xtick.color":      "#d1d5db",
        "ytick.color":      "#d1d5db",
        "grid.color":       "#374151",
        "grid.linestyle":   "--",
        "grid.alpha":       0.35,
        "font.family":      "DejaVu Sans Mono",
        "axes.titleweight": "bold",
    })

def watermark(ax, text=WATERMARK):
    ax.text(
        0.99, 0.03, text,
        transform=ax.transAxes,
        ha="right", va="bottom",
        fontsize=11, color="#9ca3af", alpha=0.35
    )

def plot_levels(df, col1, col2, title, subtitle=""):
    fig, ax = plt.subplots(figsize=(14, 6))
    ax.plot(df.index, df[col1], linewidth=1.7, label=col1)
    ax.plot(df.index, df[col2], linewidth=1.7, label=col2)
    ax.set_title(title, loc="left", fontsize=14)
    if subtitle:
        ax.text(0.01, 0.98, subtitle, transform=ax.transAxes, ha="left", va="top", fontsize=10, color="#9ca3af")
    ax.grid(True)
    ax.legend(frameon=False, loc="upper left", fontsize=10)
    watermark(ax)
    plt.tight_layout()
    plt.show()

def plot_z_with_vol(df, z_col, vol_col, low_col, title, subtitle=""):
    fig, ax1 = plt.subplots(figsize=(14, 6))

    z = df[z_col].clip(-Z_CAP, Z_CAP)
    ax1.plot(df.index, z, linewidth=1.8, label="Z-Stat (12M)")
    ax1.axhline(0, linewidth=1.0, alpha=0.8)
    ax1.axhline(2, linewidth=1.0, linestyle="--", alpha=0.8)
    ax1.axhline(-2, linewidth=1.0, linestyle="--", alpha=0.8)
    ax1.set_ylabel("Z")
    ax1.grid(True)

    ax2 = ax1.twinx()
    ax2.plot(df.index, df[vol_col], linewidth=1.2, alpha=0.85, label="Vol (12M)")
    ax2.set_ylabel("Vol")

    # low-vol shading
    low = df[low_col].fillna(False).values
    y0, y1 = ax1.get_ylim()
    ax1.fill_between(df.index, y0, y1, where=low, alpha=0.07)

    ax1.set_title(title, loc="left", fontsize=14)
    if subtitle:
        ax1.text(0.01, 0.98, subtitle, transform=ax1.transAxes, ha="left", va="top", fontsize=10, color="#9ca3af")

    # combined legend
    l1, lab1 = ax1.get_legend_handles_labels()
    l2, lab2 = ax2.get_legend_handles_labels()
    ax1.legend(l1 + l2, lab1 + lab2, frameon=False, loc="upper left", fontsize=10)

    watermark(ax1)
    plt.tight_layout()
    plt.show()

# ============================================================
# RUN EVERYTHING
# ============================================================
set_bbg_style()

# ---- metals (FRED)
copper = read_fred_monthly(F_COPPER).rename(columns={"Price": "Copper"})
iron   = read_fred_monthly(F_IRON).rename(columns={"Price": "IronOre"})

metals = copper.join(iron, how="inner").dropna()

# YoY (monthly data -> 12 lag)
metals["Copper_YoY"] = metals["Copper"] / metals["Copper"].shift(12) - 1.0
metals["Iron_YoY"]   = metals["IronOre"] / metals["IronOre"].shift(12) - 1.0

# Spread you want to test
metals["YoY_Spread"] = metals["Copper_YoY"] - metals["Iron_YoY"]

# Rolling z-stat (12m)
metals["Z_Spread_12m"] = rolling_zscore(metals["YoY_Spread"], ROLL_Z)

# Vol filter based on monthly change in spread (stability proxy)
metals["Spread_d1"] = metals["YoY_Spread"].diff()
metals = metals.join(vol_filter(metals["Spread_d1"], ROLL_VOL, BASELINE_VOL))

# ---- equities (Investing export)
china = read_investing_monthly(F_CHINA).rename(columns={"Price": "MSCI_China"})
india = read_investing_monthly(F_INDIA).rename(columns={"Price": "MSCI_India"})

asia = china.join(india, how="inner").dropna()

# Log ratio (cleaner for mean-reversion tests)
asia["LogRatio_CN_IN"] = np.log(asia["MSCI_China"] / asia["MSCI_India"])

# Rolling z-stat
asia["Z_Ratio_12m"] = rolling_zscore(asia["LogRatio_CN_IN"], ROLL_Z)

# Vol filter based on monthly change in log ratio
asia["Ratio_d1"] = asia["LogRatio_CN_IN"].diff()
asia = asia.join(vol_filter(asia["Ratio_d1"], ROLL_VOL, BASELINE_VOL))

# ============================================================
# PLOTS
# ============================================================
plot_levels(
    metals, "Copper", "IronOre",
    title="Copper vs Iron Ore (Monthly Levels)",
    subtitle="Monthly | aligned to month-start"
)

plot_z_with_vol(
    metals, z_col="Z_Spread_12m", vol_col="VOL", low_col="LOW_VOL",
    title="Copper - Iron Ore YoY Spread (Rolling 12M Z-Stat) + Vol Filter",
    subtitle="Spread = (Copper YoY - Iron Ore YoY). Shading = LOW-VOL regime."
)

plot_levels(
    asia, "MSCI_China", "MSCI_India",
    title="MSCI China vs MSCI India (Monthly Levels)",
    subtitle="Monthly | aligned to month-start"
)

plot_z_with_vol(
    asia, z_col="Z_Ratio_12m", vol_col="VOL", low_col="LOW_VOL",
    title="MSCI China / MSCI India Log Ratio (Rolling 12M Z-Stat) + Vol Filter",
    subtitle="Ratio = log(China/India). Shading = LOW-VOL regime."
)

# ============================================================
# OUTPUT TABLES (in-kernel variables, no filesystem assumptions)
# ============================================================
metals_out = metals[[
    "Copper","IronOre","Copper_YoY","Iron_YoY","YoY_Spread","Z_Spread_12m","VOL","VOL_BASE","LOW_VOL"
]].dropna()

asia_out = asia[[
    "MSCI_China","MSCI_India","LogRatio_CN_IN","Z_Ratio_12m","VOL","VOL_BASE","LOW_VOL"
]].dropna()

print("metals_out tail:")
print(metals_out.tail(5))
print("\nasia_out tail:")
print(asia_out.tail(5))

```
