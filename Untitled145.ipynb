```python
# ============================================================
# Pearl Gray Equity and Research
# DXY WEEKLY "ENTIRE PROJECT" (weekly test)
#
# INPUT: US Dollar Index Historical Data (5).csv
#
# Outputs (separate charts):
# 1) DXY Level + 52W SMA (both lines)
# 2) % Distance vs 52W SMA  = (Level / SMA - 1) * 100
# 3) Deviations (Z-score)   = (Level - SMA) / RollingStd(52W)
#
# Notes:
# - Weekly is built from the file (whatever frequency it has) -> grouped into W-FRI weeks
# - Index is relabeled to the ACTUAL last observation date in each week (no "future" week-end label confusion)
# - Bloomberg-ish dark theme + watermark + last-point labels
# ============================================================

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# ---------------------------
# CONFIG
# ---------------------------
FILE = "US Dollar Index Historical Data (5).csv"
SEARCH_ROOTS = [".", "/mnt/data", "/content"]

ROLL_W = 52
WEEKLY_RULE = "W-FRI"
WATERMARK = "Pearl Gray Equity and Research"

# ---------------------------
# STYLE (Bloomberg-ish dark)
# ---------------------------
plt.rcParams.update({
    "figure.dpi": 140,
    "savefig.dpi": 140,
    "figure.facecolor": "black",
    "axes.facecolor": "black",
    "savefig.facecolor": "black",
    "text.color": "white",
    "axes.labelcolor": "white",
    "axes.edgecolor": "white",
    "axes.titlecolor": "white",
    "xtick.color": "white",
    "ytick.color": "white",
    "font.family": "DejaVu Sans Mono",
    "axes.grid": True,
    "grid.color": "#3a3a3a",
    "grid.alpha": 0.35,
    "axes.titlesize": 12,
    "axes.labelsize": 10,
    "xtick.labelsize": 9,
    "ytick.labelsize": 9,
    "lines.linewidth": 1.8,
})

# ---------------------------
# HELPERS
# ---------------------------
def _find_path(filename: str) -> str:
    for root in SEARCH_ROOTS:
        p = os.path.join(root, filename)
        if os.path.exists(p):
            return p
    raise FileNotFoundError(f"Could not find '{filename}' in {SEARCH_ROOTS}")

def _to_float(s: pd.Series) -> pd.Series:
    s = s.astype(str).str.replace(",", "", regex=False).str.replace("%", "", regex=False).str.strip()
    s = s.replace({"": np.nan, "-": np.nan, "nan": np.nan, "NaN": np.nan})
    return pd.to_numeric(s, errors="coerce")

def load_investing_series(filename: str) -> pd.Series:
    path = _find_path(filename)
    df = pd.read_csv(path)

    if "Date" not in df.columns:
        raise ValueError(f"Missing 'Date' column. Columns: {list(df.columns)}")

    # pick value column
    val_col = "Price" if "Price" in df.columns else ("Last" if "Last" in df.columns else None)
    if val_col is None:
        val_col = [c for c in df.columns if c != "Date"][0]

    # robust date parse: YYYY/MM/DD or YYYY-MM-DD => yearfirst
    d = pd.to_datetime(df["Date"], errors="coerce", yearfirst=True, dayfirst=False)
    if d.notna().mean() < 0.8:
        # fallback if non-iso
        d = pd.to_datetime(df["Date"], errors="coerce", dayfirst=True)
        if d.notna().mean() < 0.8:
            d = pd.to_datetime(df["Date"], errors="coerce", dayfirst=False)

    v = _to_float(df[val_col])

    s = pd.Series(v.values, index=pd.DatetimeIndex(d), name="DXY").dropna().sort_index()
    return s

def to_weekly_last_actual_date(s: pd.Series, rule: str = WEEKLY_RULE) -> pd.Series:
    s = s.sort_index().dropna()
    g = s.groupby(pd.Grouper(freq=rule))

    last_vals = g.apply(lambda x: x.iloc[-1] if len(x) else np.nan).dropna()
    last_dates = g.apply(lambda x: x.index[-1] if len(x) else pd.NaT).dropna()

    last_vals.index = pd.DatetimeIndex(last_dates.values)
    return last_vals.sort_index()

def watermark(ax, text=WATERMARK):
    ax.text(
        0.5, 0.5, text,
        transform=ax.transAxes,
        fontsize=28,
        color="#bfbfbf",
        alpha=0.10,
        ha="center",
        va="center",
        rotation=25
    )

def label_last(ax, x: pd.Series, fmt="{:.3f}"):
    x = x.dropna()
    if x.empty:
        return
    t = x.index[-1]
    v = float(x.iloc[-1])
    ax.scatter([t], [v], s=28)
    ax.annotate(
        f"{t.date()}  {fmt.format(v)}",
        xy=(t, v),
        xytext=(10, 0),
        textcoords="offset points",
        ha="left",
        va="center",
        fontsize=9,
        color="white",
        bbox=dict(boxstyle="round,pad=0.25", fc="#111111", ec="#666666", alpha=0.85),
    )

def plot_two_lines(x1: pd.Series, x2: pd.Series, title: str, ylabel: str, lab1: str, lab2: str):
    fig, ax = plt.subplots(figsize=(12.6, 4.6))
    ax.plot(x1.index, x1.values, label=lab1)
    ax.plot(x2.index, x2.values, linestyle="--", label=lab2)

    ax.set_title(title)
    ax.set_ylabel(ylabel)

    ax.xaxis.set_major_locator(mdates.YearLocator(2))
    ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))

    for spine in ax.spines.values():
        spine.set_color("white")

    ax.legend(facecolor="black", edgecolor="white", labelcolor="white")
    watermark(ax)
    label_last(ax, x1)
    label_last(ax, x2)

    plt.tight_layout()
    plt.show()

def plot_one(x: pd.Series, title: str, ylabel: str, fmt="{:.3f}", zero_line=False):
    fig, ax = plt.subplots(figsize=(12.6, 4.6))
    ax.plot(x.index, x.values)

    if zero_line:
        ax.axhline(0, linewidth=1)

    ax.set_title(title)
    ax.set_ylabel(ylabel)

    ax.xaxis.set_major_locator(mdates.YearLocator(2))
    ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))

    for spine in ax.spines.values():
        spine.set_color("white")

    watermark(ax)
    label_last(ax, x, fmt=fmt)

    plt.tight_layout()
    plt.show()

# ---------------------------
# LOAD -> WEEKLY
# ---------------------------
raw = load_investing_series(FILE)
wk = to_weekly_last_actual_date(raw).rename("DXY (weekly last, actual-date index)")

start_date = wk.index.min().date()
end_date   = wk.index.max().date()
print(f"Weekly data range used: {start_date} to {end_date}")
print(f"Weekly observations: {len(wk):,}")

# ---------------------------
# CORE METRICS
# ---------------------------
sma52 = wk.rolling(ROLL_W).mean().rename("52W SMA")
sd52  = wk.rolling(ROLL_W).std(ddof=0).rename("52W Std")

pct_vs_sma = ((wk / sma52) - 1.0) * 100.0
pct_vs_sma = pct_vs_sma.rename("% vs 52W SMA")

z52 = (wk - sma52) / sd52
z52 = z52.rename("Z-score (52W)")

# ---------------------------
# PLOTS (separate)
# ---------------------------

# 1) Level + SMA
plot_two_lines(
    wk, sma52,
    title=f"DXY Weekly Level vs 52W SMA | Data: {start_date} to {end_date}",
    ylabel="Index level",
    lab1="DXY Level",
    lab2="52W SMA"
)

# 2) % distance vs SMA
plot_one(
    pct_vs_sma,
    title=f"DXY % Distance vs 52W SMA | Data: {start_date} to {end_date}",
    ylabel="% vs SMA",
    fmt="{:.2f}",
    zero_line=True
)

# 3) Deviations (Z-score)
plot_one(
    z52,
    title=f"DXY Z-score (52W) | Data: {start_date} to {end_date}",
    ylabel="z",
    fmt="{:.2f}",
    zero_line=True
)



```
