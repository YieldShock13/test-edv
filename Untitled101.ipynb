```python
# ============================================================
# Lending Rate Spreads by Loan Type
# Orca's Capital Strategies
# ============================================================

import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# Config
# -----------------------------
OCEAN_BG   = "#0b1e2d"
LINE_COLORS = ["#17becf", "#ff7f0e"]
WATERMARK  = "Orca's Capital Strategies"

# -----------------------------
# Helper: Load FRED CSVs
# -----------------------------
def load_fred_series(path, label):
    df = pd.read_csv(path)
    dcol = [c for c in df.columns if "DATE" in c.upper()][0]
    vcol = [c for c in df.columns if c != dcol][0]
    df[dcol] = pd.to_datetime(df[dcol])
    df[vcol] = pd.to_numeric(df[vcol], errors="coerce")
    df = df.set_index(dcol).rename(columns={vcol: label})
    return df

# -----------------------------
# Load Data
# -----------------------------
fedfunds = load_fred_series("FEDFUNDS (3).csv", "Fed Funds Rate")
repo     = load_fred_series("RRPONTSYAWARD.csv", "Reverse Repo Rate")
prime    = load_fred_series("DPRIME.csv", "Prime Loan Rate")
mort30   = load_fred_series("MORTGAGE30US (4).csv", "30Y Mortgage Rate")
credit   = load_fred_series("TERMCBCCALLNS.csv", "Credit Card Loan Rate")

# Merge & align
data = pd.concat([fedfunds, repo, prime, mort30, credit], axis=1).sort_index().dropna(how="any")

# -----------------------------
# Compute Spreads by Loan Type
# -----------------------------
spreads_mortgage = pd.DataFrame({
    "Mortgage  Fed Funds": data["30Y Mortgage Rate"] - data["Fed Funds Rate"],
    "Mortgage  Repo": data["30Y Mortgage Rate"] - data["Reverse Repo Rate"]
})

spreads_prime = pd.DataFrame({
    "Prime  Fed Funds": data["Prime Loan Rate"] - data["Fed Funds Rate"],
    "Prime  Repo": data["Prime Loan Rate"] - data["Reverse Repo Rate"]
})

spreads_credit = pd.DataFrame({
    "Credit Card  Fed Funds": data["Credit Card Loan Rate"] - data["Fed Funds Rate"],
    "Credit Card  Repo": data["Credit Card Loan Rate"] - data["Reverse Repo Rate"]
})

# -----------------------------
# Plot Function
# -----------------------------
def plot_spreads(df, title):
    plt.style.use("dark_background")
    fig, ax = plt.subplots(figsize=(13,6))

    for i, col in enumerate(df.columns):
        color = LINE_COLORS[i % len(LINE_COLORS)]
        ax.plot(df.index, df[col], lw=1.8, color=color, label=col)

        # Label last value (no date)
        last_date = df.index[-1]
        last_val  = df[col].iloc[-1]
        ax.scatter(last_date, last_val, color=color, s=35, zorder=5)
        ax.annotate(f"{last_val:.2f}%", xy=(last_date, last_val),
                    xytext=(6,4), textcoords="offset points",
                    fontsize=8, color=color, weight="bold")

    ax.set_facecolor(OCEAN_BG)
    ax.figure.set_facecolor(OCEAN_BG)
    ax.set_title(title, fontsize=14, color="white", weight="bold", pad=16)
    ax.set_ylabel("Spread (percentage points)", color="white", fontsize=11)
    ax.tick_params(colors="white")
    ax.grid(alpha=0.25, color="white", ls="--")
    ax.legend(facecolor=OCEAN_BG, edgecolor="none", fontsize=9)

    # Watermark
    ax.text(0.5, 0.05, WATERMARK, transform=ax.transAxes,
            fontsize=14, color="white", alpha=0.15,
            ha="center", va="center", rotation=25)

    plt.tight_layout()
    plt.show()

# -----------------------------
# Plot Charts
# -----------------------------
plot_spreads(spreads_mortgage, "Mortgage Lending Rate Spreads vs Fed Funds & Repo")
plot_spreads(spreads_prime,    "Prime Lending Rate Spreads vs Fed Funds & Repo")
plot_spreads(spreads_credit,   "Credit Card Lending Rate Spreads vs Fed Funds & Repo")




```
