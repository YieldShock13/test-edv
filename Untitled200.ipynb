```python
# ==========================================
# TREASURY vs CDS vs CREDIT ADJUSTED RATE
# + CDS standalone chart
# + 60d rolling Z-score of spread
# ==========================================

import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8-darkgrid")

plt.rcParams["figure.figsize"] = (14,7)
plt.rcParams["font.family"] = "DejaVu Sans"
plt.rcParams["axes.titlesize"] = 16


# -------------------------------
# LOAD DATA
# -------------------------------

treasury = pd.read_csv("United States 10-Year Bond Yield Historical Data (16).csv")
cds = pd.read_csv("USGV10YUSAB=R Overview (8).csv")


# -------------------------------
# CLEAN TREASURY
# -------------------------------

treasury = treasury.rename(columns={
    treasury.columns[0]: "Date",
    treasury.columns[1]: "TreasuryYield"
})

treasury["Date"] = pd.to_datetime(treasury["Date"])
treasury["TreasuryYield"] = treasury["TreasuryYield"].astype(str).str.replace("%","").astype(float)
treasury = treasury.sort_values("Date")


# -------------------------------
# CLEAN CDS
# -------------------------------

cds = cds.rename(columns={
    cds.columns[0]: "Date",
    cds.columns[1]: "CDS"
})

cds["Date"] = pd.to_datetime(cds["Date"])
cds["CDS"] = cds["CDS"].astype(float)

# convert bps → %
cds["CDS"] = cds["CDS"] / 100


# -------------------------------
# MERGE
# -------------------------------

df = pd.merge(treasury, cds, on="Date", how="inner")


# -------------------------------
# SPREAD (NOT ROLLED)
# -------------------------------

df["Spread"] = df["TreasuryYield"] - df["CDS"]


# -------------------------------
# 60d ROLLING Z-SCORE (ONLY THIS IS ROLLED)
# -------------------------------

window = 60

df["Spread_mean"] = df["Spread"].rolling(window).mean()
df["Spread_std"] = df["Spread"].rolling(window).std()

df["Spread_Z"] = (df["Spread"] - df["Spread_mean"]) / df["Spread_std"]


# ===============================
# CHART 1 — TREASURY vs SPREAD
# ===============================

fig, ax = plt.subplots()

ax.plot(df["Date"], df["TreasuryYield"], label="US 10Y Treasury", linewidth=2)
ax.plot(df["Date"], df["Spread"], label="Treasury - CDS", linewidth=3)

last = df.iloc[-1]

ax.scatter(last["Date"], last["TreasuryYield"])
ax.text(last["Date"], last["TreasuryYield"], f"{last['TreasuryYield']:.2f}")

ax.scatter(last["Date"], last["Spread"])
ax.text(last["Date"], last["Spread"], f"{last['Spread']:.2f}")

ax.set_title("Treasury vs Credit-Adjusted Rate")
ax.legend()

plt.show()



# ===============================
# CHART 2 — CDS ONLY
# ===============================

fig, ax = plt.subplots()

ax.plot(df["Date"], df["CDS"], color="orange", linewidth=2, label="US Sovereign CDS")

last = df.iloc[-1]

ax.scatter(last["Date"], last["CDS"])
ax.text(last["Date"], last["CDS"], f"{last['CDS']:.2f}")

ax.set_title("US Sovereign CDS (10Y)")
ax.legend()

plt.show()



# ===============================
# CHART 3 — 60D Z-SCORE
# ===============================

fig, ax = plt.subplots()

ax.plot(df["Date"], df["Spread_Z"], color="purple", linewidth=2, label="Spread Z-score (60d)")

ax.axhline(2, linestyle="--")
ax.axhline(-2, linestyle="--")
ax.axhline(0)

last = df.iloc[-1]

ax.scatter(last["Date"], last["Spread_Z"])
ax.text(last["Date"], last["Spread_Z"], f"{last['Spread_Z']:.2f}")

ax.set_title("60-Day Rolling Z-Score of (Treasury - CDS)")
ax.legend()

plt.show()

```
