```python
# ===============================
# GERMAN CURVE PCA - CLEAN VERSION
# ===============================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

plt.style.use("seaborn-v0_8-darkgrid")

# -------------------------------
# LOAD DATA
# -------------------------------

y2 = pd.read_csv("Germany 2-Year Bond Yield Historical Data (4).csv")
y5 = pd.read_csv("Germany 5-Year Bond Yield Historical Data.csv")
y10 = pd.read_csv("Germany 10-Year Bond Yield Historical Data (4).csv")

def clean(df, name):
    df = df.rename(columns={df.columns[0]: "Date", df.columns[1]: name})
    df["Date"] = pd.to_datetime(df["Date"])
    df[name] = (
        df[name].astype(str)
        .str.replace("%","", regex=False)
        .str.replace(",","", regex=False)
        .astype(float)
    )
    return df[["Date", name]]

y2 = clean(y2,"Y2")
y5 = clean(y5,"Y5")
y10 = clean(y10,"Y10")

df = y2.merge(y5,on="Date").merge(y10,on="Date")
df = df.sort_values("Date")

# -------------------------------
# PCA INPUT = DAILY CHANGES
# -------------------------------

X = df[["Y2","Y5","Y10"]].diff().dropna()

# -------------------------------
# PCA
# -------------------------------

pca = PCA(n_components=3)
scores = pca.fit_transform(X)

loadings = pd.DataFrame(
    pca.components_.T,
    index=["2Y","5Y","10Y"],
    columns=["PC1","PC2","PC3"]
)

variance = pca.explained_variance_ratio_

print("Explained variance:", variance)
print(loadings)

# ===============================
# CHART 1 — VARIANCE EXPLAINED
# ===============================

plt.figure(figsize=(8,5))
plt.bar(["PC1","PC2","PC3"], variance*100)
plt.ylabel("Percent variance explained")
plt.title("Yield Curve PCA – Variance Explained")
plt.show()

# ===============================
# CHART 2 — FACTOR LOADINGS
# ===============================

plt.figure(figsize=(10,6))
for col in loadings.columns:
    plt.plot(loadings.index, loadings[col], marker="o", label=col)

plt.axhline(0,color="black",linewidth=1)
plt.title("Yield Curve PCA – Factor Loadings")
plt.ylabel("Loading")
plt.legend()
plt.show()

# ===============================
# CHART 3 — FACTOR SCORES
# ===============================

scores_df = pd.DataFrame(
    scores,
    columns=["PC1","PC2","PC3"],
    index=X.index
)

fig, axes = plt.subplots(3,1,figsize=(12,10),sharex=True)

for i,col in enumerate(scores_df.columns):
    axes[i].plot(df["Date"].iloc[1:], scores_df[col])
    axes[i].set_title(col)

plt.tight_layout()
plt.show()
```
