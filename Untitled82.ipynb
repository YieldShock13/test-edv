```python
# ==========================================================
# Confluence Strategy (Monthly 1st) — β-hedged Long–Short
# Signals (confluence):
#   LONG US / SHORT DE when  (Z_yield <= -THRESH) & (Z_fut >= +THRESH)
#   LONG DE / SHORT US when  (Z_yield >= +THRESH) & (Z_fut <= -THRESH)
# Else FLAT
#
# β from: DE_fut ~ const + β * US_fut
# Pair return normalized by gross (β+1) for consistent sizing.
#
# Thresholds: {0.0, 0.25, 0.5, 1.0}
# Leverages:  {100x, 200x, 300x}
# Slippage:   fixed 4 bps RT per active month (NOT scaled by leverage)
#
# Plots: z-scores, entries, equity (unlev + levered), drawdowns, monthly bars
# CSV:   per-threshold outputs
#
# Watermark: "Orca's Capital Strategies"
# Start date: 2015-01-01
# ==========================================================

import os, glob, warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import statsmodels.api as sm

warnings.filterwarnings("ignore", category=UserWarning)

try:
    from google.colab import files as colab_files
    _COLAB = True
except Exception:
    _COLAB = False

# ---------------- Config ----------------
WATERMARK = "Orca's Capital Strategies"
DATE_FMT_M = DateFormatter("%Y-%m")
THRESH_LIST = [0.0, 0.25, 0.5, 1.0]
LEVERAGE_LIST = [100.0, 200.0, 300.0]

# Latest filenames you uploaded
FN_US_YLD = "United States 2-Year Bond Yield Historical Data (2).csv"
FN_DE_YLD = "Germany 2-Year Bond Yield Historical Data (3).csv"
FN_US_FUT = "US 2 Year T-Note Futures Historical Data (1).csv"
FN_DE_FUT = "Euro SCHATZ Futures Historical Data (1).csv"

SLIPPAGE_BPS_PER_ROUNDTRIP = 4       # fixed 4 bps per active month
START_EQUITY = 100_000.0
PERIODS_PER_YEAR = 12
START_DATE = "2015-01-01"

# ---------------- Helpers ----------------
def resolve_path(filename):
    for p in (f"/content/{filename}", f"/mnt/data/{filename}", filename):
        if os.path.isfile(p): return p
    for base in ("/content", "/mnt/data", "."):
        hits = glob.glob(os.path.join(base, "**", filename), recursive=True)
        if hits: return hits[0]
    raise FileNotFoundError(f"Could not find: {filename}")

def robust_date(s: pd.Series) -> pd.Series:
    # Files are generally US-style; prefer dayfirst=False
    return pd.to_datetime(s, errors="coerce", dayfirst=False)

def parse_number(series: pd.Series) -> pd.Series:
    s = series.astype(str).str.strip().str.replace("%","",regex=False)
    if len(s) == 0:
        return pd.to_numeric(s, errors="coerce")
    smpl = s.sample(min(100, len(s)), random_state=0)
    comma_no_dot = ((smpl.str.contains(",")) & ~smpl.str.contains(r"\.")).mean()
    if comma_no_dot > 0.6:
        s = s.str.replace(".", "", regex=False)  # thousands dot
        s = s.str.replace(",", ".", regex=False) # decimal comma -> dot
    else:
        s = s.str.replace(",", "", regex=False)  # thousands comma
    return pd.to_numeric(s, errors="coerce")

def load_monthly(path: str, colname: str) -> pd.Series:
    df = pd.read_csv(path)
    df.columns = [c.strip() for c in df.columns]
    if "Date" not in df.columns or "Price" not in df.columns:
        raise ValueError(f"'Date' and 'Price' required in {path}. Found: {df.columns.tolist()}")
    df["Date"]  = robust_date(df["Date"])
    df["Price"] = parse_number(df["Price"])
    df = df.dropna(subset=["Date","Price"]).sort_values("Date").set_index("Date")
    # Monthly 1st-of-month
    df = df.resample("MS").first()
    return df["Price"].rename(colname)

def add_wm(fig): fig.text(0.01, 0.01, WATERMARK, fontsize=12, alpha=0.35, ha="left", va="bottom")

def annotate_last(ax, x, y, fmt="{:.2f}", dx=8, dy=0):
    if len(y) > 0:
        ax.annotate(fmt.format(y[-1]), xy=(x[-1], y[-1]), xytext=(dx, dy),
                    textcoords="offset points", va="center")

def finalize(fig):
    add_wm(fig); plt.tight_layout(); plt.show()

def equity_from_returns(inc: pd.Series, start=1.0) -> pd.Series:
    inc = inc.fillna(0.0)
    return pd.Series(start, index=inc.index).mul((1 + inc).cumprod())

def max_drawdown(eq: pd.Series):
    roll = eq.cummax()
    dd = eq/roll - 1.0
    return dd.min(), dd

def safe_cagr(inc: pd.Series, periods_per_year=12):
    inc = inc.dropna()
    if inc.empty:
        return np.nan
    total = (1 + inc).prod()
    if total <= 0:
        return np.nan
    years = len(inc) / periods_per_year
    if years <= 0:
        return np.nan
    return total**(1/years) - 1

# ---------------- Load ----------------
US2Y = load_monthly(resolve_path(FN_US_YLD), "US2Y_yld")
DE2Y = load_monthly(resolve_path(FN_DE_YLD), "DE2Y_yld")
US_F = load_monthly(resolve_path(FN_US_FUT), "US_fut")
DE_F = load_monthly(resolve_path(FN_DE_FUT), "DE_fut")

US2Y, DE2Y, US_F, DE_F = [s.loc[s.index >= START_DATE] for s in (US2Y, DE2Y, US_F, DE_F)]

yld = pd.concat([DE2Y, US2Y], axis=1).dropna()
fut = pd.concat([DE_F, US_F], axis=1).dropna()

# ---------------- Z-scores ----------------
# Yield spread (bps) z
yld["Spread_bps"] = (yld["DE2Y_yld"] - yld["US2Y_yld"]) * 100.0
mu_y, sd_y = yld["Spread_bps"].mean(), yld["Spread_bps"].std(ddof=0)
yld["Z_yield"] = (yld["Spread_bps"] - mu_y) / sd_y if sd_y != 0 else np.nan

# Futures hedge β via OLS: DE_fut ~ const + β·US_fut
Xf = sm.add_constant(fut["US_fut"])
yf = fut["DE_fut"]
ols_fut = sm.OLS(yf, Xf).fit()
beta_f = float(ols_fut.params["US_fut"])
alpha_f = float(ols_fut.params.get("const", 0.0))

# Hedged spread z: (DE - β·US)
fut["Spread_hedged"] = fut["DE_fut"] - beta_f * fut["US_fut"]
mu_h, sd_h = fut["Spread_hedged"].mean(), fut["Spread_hedged"].std(ddof=0)
fut["Z_fut"] = (fut["Spread_hedged"] - mu_h) / sd_h if sd_h != 0 else np.nan

# ---------------- Forward returns (1M ahead) ----------------
DE_ret = DE_F.shift(-1) / DE_F - 1.0
US_ret = US_F.shift(-1) / US_F - 1.0
ret_df = pd.concat([DE_ret.rename("DE_ret"), US_ret.rename("US_ret")], axis=1)

# Strategy panel
panel = fut[["Z_fut"]].join(yld[["Z_yield"]], how="inner").join(ret_df, how="inner").dropna()

# ---------------- Metrics helpers ----------------
def eval_leverage(unlev_ret: pd.Series, active: pd.Series, leverage: float):
    # Fixed hairline slippage (NOT scaled by leverage)
    slip_frac = SLIPPAGE_BPS_PER_ROUNDTRIP / 10_000.0
    slip_series = active.astype(float) * slip_frac

    lev_net = unlev_ret * leverage - slip_series
    eq_lev = equity_from_returns(lev_net)
    mdd, dd = max_drawdown(eq_lev)
    ann_vol = lev_net.std(ddof=0) * np.sqrt(PERIODS_PER_YEAR)
    var95 = -np.nanpercentile(lev_net.dropna(), 5) if lev_net.notna().sum() > 0 else np.nan
    return {
        "ret": lev_net, "eq": eq_lev, "dd": dd, "mdd": mdd,
        "cum_gain": eq_lev.iloc[-1]-1.0 if len(eq_lev)>0 else np.nan,
        "cagr": safe_cagr(lev_net), "ann_vol": ann_vol, "var95": var95,
        "slip_series": slip_series
    }

def compute_unlevered(unlev_ret: pd.Series):
    eq = equity_from_returns(unlev_ret)
    mdd, dd = max_drawdown(eq)
    ann_vol = unlev_ret.std(ddof=0) * np.sqrt(PERIODS_PER_YEAR)
    var95 = -np.nanpercentile(unlev_ret.dropna(), 5) if unlev_ret.notna().sum() > 0 else np.nan
    return {
        "ret": unlev_ret, "eq": eq, "dd": dd, "mdd": mdd,
        "cum_gain": eq.iloc[-1]-1.0 if len(eq)>0 else np.nan,
        "cagr": safe_cagr(unlev_ret), "ann_vol": ann_vol, "var95": var95
    }

# ---------------- Strategy Runner (β-hedged long–short) ----------------
def run_confluence_longshort(thresh: float):
    """
    LONG US / SHORT DE when  (Z_yield <= -thresh) & (Z_fut >= +thresh)
    LONG DE / SHORT US when  (Z_yield >= +thresh) & (Z_fut <= -thresh)
    Else FLAT
    Pair return uses β hedge and is normalized by gross (β+1).
    """
    df = panel.copy()

    # Signals
    df["sig_USlong"] = (df["Z_yield"] <= -thresh) & (df["Z_fut"] >= +thresh)
    df["sig_DElong"] = (df["Z_yield"] >= +thresh) & (df["Z_fut"] <= -thresh)
    df["active"]     = df["sig_USlong"] | df["sig_DElong"]

    # β-hedged, gross-normalized pair returns (held to next month)
    gross = beta_f + 1.0
    pair_USlong = (beta_f * df["US_ret"] - df["DE_ret"]) / gross     # +β·US, −1·DE
    pair_DElong = (df["DE_ret"] - beta_f * df["US_ret"]) / gross     # +1·DE, −β·US

    strat = pd.Series(0.0, index=df.index, name="Strat_ret")
    strat[df["sig_USlong"]] = pair_USlong.loc[df["sig_USlong"]]
    strat[df["sig_DElong"]] = pair_DElong.loc[df["sig_DElong"]]
    df["Strat_ret"] = strat

    # ---- Unlevered ----
    un = compute_unlevered(df["Strat_ret"])

    # ---- Levered (fixed slippage, NOT scaled) ----
    lev_results = {int(L): eval_leverage(df["Strat_ret"], df["active"], L) for L in LEVERAGE_LIST}

    # ---- Plots ----
    # 1) Z-scores
    figZ, axZ = plt.subplots(figsize=(11.2, 5.2))
    zy = yld.reindex(df.index)["Z_yield"]; zf = fut.reindex(df.index)["Z_fut"]
    axZ.plot(df.index, zy, label="Yield z (DE−US, bps)")
    axZ.plot(df.index, zf, label=f"Futures z (DE − β·US), β={beta_f:.3f}")
    axZ.axhline(0, linewidth=1)
    if thresh > 0:
        axZ.axhline(+thresh, linestyle="--", alpha=0.8, label=f"+{thresh}σ")
        axZ.axhline(-thresh, linestyle="--", alpha=0.8, label=f"−{thresh}σ")
    axZ.set_title(f"Z-Scores — Confluence (threshold = {thresh}σ)")
    axZ.set_ylabel("Z"); axZ.set_xlabel("Date"); axZ.grid(True, alpha=0.3)
    axZ.xaxis.set_major_formatter(DATE_FMT_M)
    annotate_last(axZ, df.index, zy.values, fmt="{:.2f}")
    annotate_last(axZ, df.index, zf.values, fmt="{:.2f}", dy=-12)
    axZ.legend(); finalize(figZ)

    # 2) Entry markers on futures (context)
    figP, axP = plt.subplots(figsize=(12.0, 5.8))
    axP.plot(fut.index, fut["DE_fut"], label="DE futures")
    axP.plot(fut.index, fut["US_fut"], label="US futures")
    axP.scatter(df.index[df["sig_DElong"]], fut.reindex(df.index[df["sig_DElong"]])["DE_fut"],
                marker="^", s=90, label="ENTRY: Long DE / Short US", zorder=5)
    axP.scatter(df.index[df["sig_USlong"]], fut.reindex(df.index[df["sig_USlong"]])["US_fut"],
                marker="v", s=90, label="ENTRY: Long US / Short DE", zorder=5)
    axP.set_title(f"Entries — β-hedged Long–Short (threshold = {thresh}σ)")
    axP.set_ylabel("Price"); axP.set_xlabel("Date"); axP.grid(True, alpha=0.3)
    axP.xaxis.set_major_formatter(DATE_FMT_M)
    annotate_last(axP, fut.index, fut["DE_fut"].values, fmt="{:.2f}")
    annotate_last(axP, fut.index, fut["US_fut"].values, fmt="{:.2f}")
    axP.legend(); finalize(figP)

    # 3) Equity curves: unlevered + levered
    figE, axE = plt.subplots(figsize=(11.6, 5.2))
    axE.plot(df.index, un["eq"], label="Equity (Unlevered)")
    for L in LEVERAGE_LIST:
        axE.plot(df.index, lev_results[int(L)]["eq"], label=f"Equity (Levered {int(L)}x, net)")
    axE.set_title(f"Equity — β-hedged Long–Short (threshold = {thresh}σ)")
    axE.set_ylabel("Growth of $1"); axE.set_xlabel("Date"); axE.grid(True, alpha=0.3)
    axE.xaxis.set_major_formatter(DATE_FMT_M)
    axE.legend(); finalize(figE)

    # 4) Monthly return bars (unlevered)
    figR, axR = plt.subplots(figsize=(11.2, 4.2))
    colors = np.where(df["Strat_ret"] >= 0, "tab:blue", "tab:orange")
    axR.bar(df.index, df["Strat_ret"].values, width=20, align="center", color=colors)
    axR.axhline(0, linewidth=1)
    axR.set_title(f"Monthly Returns (Unlevered) — threshold {thresh}σ")
    axR.set_ylabel("Return"); axR.set_xlabel("Date"); axR.grid(True, alpha=0.25)
    axR.xaxis.set_major_formatter(DATE_FMT_M)
    finalize(figR)

    # 5) Drawdowns: unlevered + levered
    figD, axD = plt.subplots(figsize=(11.2, 3.8))
    axD.plot(df.index, un["dd"], label="Drawdown (Unlevered)")
    for L in LEVERAGE_LIST:
        axD.plot(df.index, lev_results[int(L)]["dd"], label=f"Drawdown (Levered {int(L)}x)")
    axD.set_title(f"Drawdown — β-hedged Long–Short (threshold = {thresh}σ)")
    axD.set_ylabel("Drawdown"); axD.set_xlabel("Date"); axD.grid(True, alpha=0.3)
    axD.xaxis.set_major_formatter(DATE_FMT_M)
    axD.legend(); finalize(figD)

    # ---- Print metrics ----
    print(f"\n=== β-hedged Long–Short — threshold {thresh}σ (monthly) ===")
    print(f"β (hedge): {beta_f:.6f}  |  α: {alpha_f:.6f}")
    print(f"Months: {len(df)}  |  Active months: {int(df['active'].sum())}")
    print("--- Unlevered ---")
    print(f"CumGain: {un['cum_gain']:.2%}  |  MaxDD: {un['mdd']:.2%}  |  VaR95: {un['var95']:.2%}  |  "
          f"Ann.Vol: {un['ann_vol']:.2%}  |  CAGR: {un['cagr']:.2%}")
    for L in LEVERAGE_LIST:
        r = lev_results[int(L)]
        print(f"--- Levered ({int(L)}x, slippage {SLIPPAGE_BPS_PER_ROUNDTRIP} bps RT / active month, not scaled) ---")
        print(f"CumGain: {r['cum_gain']:.2%}  |  MaxDD: {r['mdd']:.2%}  |  VaR95: {r['var95']:.2%}  |  "
              f"Ann.Vol: {r['ann_vol']:.2%}  |  CAGR: {r['cagr']:.2%}")

    # ---- Save outputs (per threshold) ----
    out_cols = [
        yld.reindex(df.index)[["Spread_bps","Z_yield"]],
        fut.reindex(df.index)[["Spread_hedged","Z_fut"]],
        df[["DE_ret","US_ret","Strat_ret","active","sig_USlong","sig_DElong"]],
        un["eq"].rename("Equity_Unlevered"),
        un["dd"].rename("Drawdown_Unlevered"),
    ]
    for L in LEVERAGE_LIST:
        r = lev_results[int(L)]
        out_cols += [
            r["ret"].rename(f"Ret_Levered_{int(L)}x"),
            r["eq"].rename(f"Equity_Levered_{int(L)}x"),
            r["dd"].rename(f"Drawdown_Levered_{int(L)}x"),
            r["slip_series"].rename(f"SlippageFrac_{int(L)}x"),
        ]
    out = pd.concat(out_cols, axis=1)

    out_path = f"/content/confluence_longshort_thresh_{str(thresh).replace('.','p')}.csv"
    out.to_csv(out_path)
    print(f"Saved CSV -> {out_path}")
    if _COLAB:
        try:
            colab_files.download(out_path)
        except Exception:
            pass

# ---------------- RUN ----------------
print("Confluence (β-hedged long–short) confirmation:")
print(" Long US / Short DE when (Z_yield <= -THRESH) & (Z_fut >= +THRESH)")
print(" Long DE / Short US when (Z_yield >= +THRESH) & (Z_fut <= -THRESH)")
print(" Else FLAT. Rebalance monthly on the 1st; hold 1 month.")

for thr in THRESH_LIST:
    run_confluence_longshort(thr)





```
