```python
# ============================================================
# Gold–Silver Futures Pair — AND rules, new-trigger, 1M hold
# VIX (12M Z, 3M MA) & Ratio (3M Z); thresholds: 1.00, 0.90, 0.75, 0.50
# Beta-hedged (24M OLS, shrink+clip); Vol target (15% on unit core, cap 3x)
# BASE_LEVERAGE in {10x, 20x}; Slippage 4bps per 1x notional (scaled)
# Upside-only winsor at 95% (ACTIVE leg only); Cash yield layer added monthly
# Metrics on FULL monthly series (cash + active; active=0 when flat)
# Files (monthly, Investing.com):
#   - "Gold Futures Historical Data (2).csv"
#   - "Silver Futures Historical Data.csv"
#   - "CBOE Volatility Index Historical Data (2).csv"
# ============================================================

import pandas as pd, numpy as np, matplotlib.pyplot as plt
plt.style.use("dark_background")
BG="#0b1e2d"; GREEN="#2ca02c"; RED="#d62728"; CYAN="#17becf"; AMBER="#f2c744"

# ---------------- User knobs ----------------
THRESHOLDS = [1.00, 0.90, 0.75, 0.50]   # sigma levels for BOTH rules
BASE_LEVERAGES = [10.0, 20.0]           # run both 10x and 20x
ROLL_VIX, SMOOTH_VIX = 12, 3            # VIX Z window, then 3M smoothing
ROLL_RATIO = 3                           # ratio Z window (short-term)
ROLL_BETA  = 24                          # OLS hedge window
ROLL_VOL   = 12                          # realized vol lookback for sizing
TARGET_VOL = 0.15                        # target ann vol for unit-notional core
SIZE_MAX   = 3.0                         # cap on size from vol targeting
SLIPPAGE_PER_1X = 0.0004                # 4 bps per 1× notional, per executed month (log space)
WINSOR_UP = 95                           # upside-only winsor (percentile)
CASH_YIELD_ANNUAL = 0.03                 # 3% annual yield on collateral (T-bills, RRP, etc.)
CASH_EVERY_MONTH = True                  # earn cash yield in *all* months (including while invested)

# ---------------- Data loaders ----------------
def load_investing_monthly(fn):
    df = pd.read_csv(fn)
    df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
    df = df.dropna(subset=["Date"]).sort_values("Date")
    df["Price"] = df["Price"].astype(str).str.replace(",", "", regex=False).astype(float)
    return df.set_index("Date")["Price"]

gold_p   = load_investing_monthly("Gold Futures Historical Data (2).csv")
silver_p = load_investing_monthly("Silver Futures Historical Data.csv")
vix_p    = load_investing_monthly("CBOE Volatility Index Historical Data (2).csv")

# Align & log returns
idx = gold_p.index.union(silver_p.index).union(vix_p.index)
gold_p, silver_p, vix_p = gold_p.reindex(idx).ffill(), silver_p.reindex(idx).ffill(), vix_p.reindex(idx).ffill()
gold_r, silver_r = np.log(gold_p/gold_p.shift(1)), np.log(silver_p/silver_p.shift(1))

# Signals
vix_z_raw = (vix_p - vix_p.rolling(ROLL_VIX).mean()) / vix_p.rolling(ROLL_VIX).std(ddof=0)
vix_z     = vix_z_raw.rolling(SMOOTH_VIX).mean()
ratio     = gold_p / silver_p
ratio_z   = (ratio - ratio.rolling(ROLL_RATIO).mean()) / ratio.rolling(ROLL_RATIO).std(ddof=0)

# Common index after burn-in
midx = gold_r.index.intersection(silver_r.index).intersection(vix_z.dropna().index).intersection(ratio_z.dropna().index)
gold_r, silver_r, vix_z, ratio_z = gold_r.reindex(midx), silver_r.reindex(midx), vix_z.reindex(midx), ratio_z.reindex(midx)

# ---------------- Hedge betas (stabilized) ----------------
def rolling_beta(y, x, win=24):
    return y.rolling(win).cov(x) / x.rolling(win).var()

def stabilize_beta(b):
    b = 0.7*b + 0.3*1.0          # shrink toward 1.0
    return b.clip(lower=0.5, upper=1.5)

beta_g_on_s = stabilize_beta(rolling_beta(gold_r,  silver_r, ROLL_BETA))
beta_s_on_g = stabilize_beta(rolling_beta(silver_r, gold_r,  ROLL_BETA))

# ---------------- Cash yield (monthly, log) ----------------
# Use effective monthly growth from annual rate; then take log return.
# This approximates fully collateralized futures (collateral earns yield).
rcash_log = np.log((1.0 + CASH_YIELD_ANNUAL)**(1.0/12.0))

# ---------------- Metrics (FULL series aware) ----------------
def cagr_full(r_full_monthly):
    r = r_full_monthly.fillna(0.0)   # includes cash layer every month
    n = len(r)
    if n == 0: return np.nan
    yrs = n/12
    return (np.exp(r.sum())**(1/yrs)) - 1

def ann_vol_full(r_full_monthly):
    r = r_full_monthly.fillna(0.0)
    if r.empty: return np.nan
    return r.std(ddof=0)*np.sqrt(12)

def dd_min_full(r_full_monthly):
    r = r_full_monthly.fillna(0.0)
    if r.empty: return np.nan
    eq = np.exp(r.cumsum())
    return (eq/eq.cummax() - 1).min()

def calmar(cagr, dd):
    return np.nan if pd.isna(cagr) or pd.isna(dd) or dd == 0 else cagr/abs(dd)

def fmt_pct(x): return "—" if pd.isna(x) else f"{x*100:.2f}%"
def fmt_ratio(x): return "—" if pd.isna(x) else f"{x:.2f}"

def winsorize_upside(r_active, pct=WINSOR_UP):
    """Clip ONLY the positive tail above the given percentile, applied AFTER costs on ACTIVE months."""
    data = r_active.dropna()
    if data.empty: return r_active
    hi = np.nanpercentile(data, pct)
    return r_active.clip(upper=hi)

# ---------------- Build AND signal, new trigger, 1M hold ----------------
def build_direction(thr):
    # -1 = long gold/short silver, +1 = long silver/short gold, 0 = flat
    dir_t = pd.Series(0, index=midx, dtype=int)
    dir_t[(vix_z <= -thr) & (ratio_z <= -thr)] = -1
    dir_t[(vix_z >=  thr) & (ratio_z >=  thr)] = +1
    # new trigger only (neutral -> signal)
    new_entry = (dir_t != 0) & (dir_t.shift(1).fillna(0) == 0)
    # execute next month for 1M
    exec_mask = new_entry.shift(1).fillna(False)
    lg_exec = exec_mask & (dir_t.shift(1) == -1)
    ls_exec = exec_mask & (dir_t.shift(1) == +1)
    return dir_t, lg_exec, ls_exec

# ---------------- Core backtest for one (leverage, threshold) ----------------
def run_case(base_leverage, thr):
    dir_t, lg_exec, ls_exec = build_direction(thr)

    # betas to use at t+1
    b_g_next = beta_g_on_s.shift(1)
    b_s_next = beta_s_on_g.shift(1)

    # unit-notional core (beta-hedged) on ACTIVE months only
    core_unit = pd.Series(np.nan, index=midx)
    if lg_exec.any():
        beta = b_g_next[lg_exec].replace([np.inf,-np.inf],np.nan).fillna(1.0)
        core_unit.loc[lg_exec] = gold_r[lg_exec] - beta*silver_r[lg_exec]
    if ls_exec.any():
        beta = b_s_next[ls_exec].replace([np.inf,-np.inf],np.nan).fillna(1.0)
        core_unit.loc[ls_exec] = silver_r[ls_exec] - beta*gold_r[ls_exec]

    # vol targeting on unit-notional core (use history up to t)
    hist_core = core_unit.shift(1)
    roll_vol = hist_core.rolling(ROLL_VOL).std(ddof=0) * np.sqrt(12)
    size = pd.Series(1.0, index=midx)
    mask = (roll_vol>0) & ~np.isnan(roll_vol)
    size.loc[mask] = np.minimum(SIZE_MAX, TARGET_VOL/roll_vol.loc[mask])

    # apply base leverage
    eff_size = size * base_leverage

    # post-slippage net ACTIVE (log), only on active months
    net_active = pd.Series(np.nan, index=midx)
    active = core_unit.notna()
    if active.any():
        net_active.loc[active] = eff_size.loc[active]*core_unit.loc[active] - eff_size.loc[active]*SLIPPAGE_PER_1X

    # UPSIDE-ONLY WINSOR on ACTIVE months
    net_active = winsorize_upside(net_active, WINSOR_UP)

    # Cash layer: add rcash_log to EVERY month (including when invested)
    cash_series = pd.Series(rcash_log, index=midx) if CASH_EVERY_MONTH else pd.Series(0.0, index=midx)

    # FULL monthly series = cash + active (active = 0 when flat)
    net_full = cash_series + net_active.fillna(0.0)

    # ------- Stats -------
    trades = int(net_active.dropna().shape[0])
    win    = (net_active.dropna() > 0).mean() if trades>0 else np.nan

    cagr   = cagr_full(net_full)
    vol    = ann_vol_full(net_full)
    dd     = dd_min_full(net_full)
    cal    = calmar(cagr, dd)
    total  = np.exp(net_full.cumsum()).iloc[-1] - 1
    trr    = (cagr/vol) if (not pd.isna(cagr) and not pd.isna(vol) and vol!=0) else np.nan

    # Ledger (active months only) — include cash addendum for context
    rows=[]
    for dt in net_active.dropna().index:
        if lg_exec.loc[dt]:
            dsc="Long Gold / Short Silver"; hb=b_g_next.loc[dt]
        elif ls_exec.loc[dt]:
            dsc="Long Silver / Short Gold"; hb=b_s_next.loc[dt]
        else:
            dsc=""; hb=np.nan
        rows.append([dt - pd.offsets.MonthEnd(1), dt, dsc,
                     float(eff_size.loc[dt]) if pd.notna(eff_size.loc[dt]) else np.nan,
                     float(hb) if pd.notna(hb) else np.nan,
                     float((net_active.loc[dt])*100),
                     float((rcash_log)*100),
                     float((net_full.loc[dt])*100)])
    ledger = pd.DataFrame(rows, columns=[
        "SignalMonth","TradeMonth","Direction","Size","HedgeBeta",
        "ActiveNet_%","Cash_%","TotalNet_%"
    ])

    stats = {"Lev":base_leverage,"Thr":thr,"Trades":trades,"WinRate":win,"CAGR":cagr,
             "AnnVol":vol,"MaxDD":dd,"CalMAR":cal,"Total":total,"TotalRiskRatio":trr}
    return net_active, net_full, ledger, stats

# ---------------- Run all cases ----------------
all_stats=[]; all_ledgers={}
all_series_full={}  # for plotting full-series equity/DD
for lev in BASE_LEVERAGES:
    for thr in THRESHOLDS:
        net_active, net_full, led, st = run_case(lev, thr)
        all_stats.append(st); all_ledgers[(lev,thr)] = (net_active, net_full, led)
        all_series_full[(lev,thr)] = net_full

summary = pd.DataFrame(all_stats)[["Lev","Thr","Trades","WinRate","CAGR","AnnVol","MaxDD","CalMAR","Total","TotalRiskRatio"]]
for c in ["WinRate","CAGR","AnnVol","MaxDD","Total"]:
    summary[c] = summary[c].apply(fmt_pct)
summary["CalMAR"] = summary["CalMAR"].apply(fmt_ratio)
summary["TotalRiskRatio"] = summary["TotalRiskRatio"].apply(fmt_ratio)

print("\n=== Gold–Silver Futures — AND rules (new-trigger, 1M hold) ===")
print("Post-slippage (6bps/1x), 95% upside-winsor on ACTIVE leg, + 3%/yr cash yield every month")
print("CAGR/Vol/DD/CalMAR/Total computed on FULL series (cash + active; active=0 when flat).")
print("WinRate = trade months only.\n")
print(summary.to_string(index=False))

# ---------------- Charts per (leverage, threshold) — FULL series ----------------
def draw_case(lev, thr, net_full, net_active):
    if net_full.empty:
        print(f"\nNo data for Lev={lev:.0f}x Thr={thr:.2f}."); return
    tag=f"Lev {lev:.0f}× | Thr {thr:.2f}σ | 95% winsor (active) | 6bps/1x | 3%/yr cash"
    net_pct_active = net_active.dropna()*100

    fig, ax = plt.subplots(3,1,figsize=(12,12),sharex=True)

    # 1) Monthly net returns — ACTIVE leg only (post-cost, post-winsor)
    if not net_pct_active.empty:
        ax[0].bar(net_pct_active.index, net_pct_active.values, width=25,
                  color=np.where(net_pct_active.values>=0, GREEN, RED), alpha=0.85)
    ax[0].axhline(0, color="white", lw=1)
    ax[0].set_facecolor(BG)
    ax[0].set_title(f"Monthly Net Returns (ACTIVE only) — {tag}", color="white", fontsize=13)
    ax[0].set_ylabel("%", color="white"); ax[0].tick_params(colors="white")
    ax[0].grid(alpha=0.3, color="white", ls="--")

    # 2) Equity (FULL series: cash + active)
    eq=np.exp(net_full.fillna(0).cumsum())
    ax[1].plot(eq.index, eq.values, color=CYAN, lw=1.8)
    ax[1].set_facecolor(BG)
    ax[1].set_title("Cumulative Growth of $1 — FULL series (cash + active)", color="white", fontsize=13)
    ax[1].set_ylabel("×", color="white"); ax[1].tick_params(colors="white")
    ax[1].grid(alpha=0.3, color="white", ls="--")

    # 3) Drawdown (FULL series)
    dd = eq/eq.cummax() - 1
    ax[2].plot(dd.index, dd.values, lw=1.5)
    ax[2].axhline(0, color="white", lw=1)
    ax[2].set_facecolor(BG)
    ax[2].set_title("Drawdown — FULL series (cash + active)", color="white", fontsize=13)
    ax[2].set_ylabel("DD", color="white"); ax[2].tick_params(colors="white")
    ax[2].grid(alpha=0.3, color="white", ls="--")

    plt.tight_layout(); plt.show()

for (lev, thr), (net_active, net_full, led) in all_ledgers.items():
    draw_case(lev, thr, net_full, net_active)

# ---------------- (Optional) peek at ledgers ----------------
for (lev, thr), (net_active, net_full, led) in all_ledgers.items():
    print(f"\n--- Ledger sample | Lev {lev:.0f}×, Thr {thr:.2f}σ ---")
    print(led.head(12).to_string(index=False))

_ = None


_ = None
```
