```python
# Install dependencies
!pip install -q pandas numpy matplotlib requests

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests

# API key
api_key = "804d445164667f91a836df826235e5cc"

# Fetch FRED data
def fetch_fred_series(series_id, start="2000-01-01", end="2025-07-01"):
    url = f"https://api.stlouisfed.org/fred/series/observations"
    params = {
        "series_id": series_id,
        "api_key": api_key,
        "file_type": "json",
        "observation_start": start,
        "observation_end": end,
        "frequency": "m"
    }
    r = requests.get(url, params=params)
    data = r.json()['observations']
    df = pd.DataFrame(data)
    df['date'] = pd.to_datetime(df['date'])
    df.set_index('date', inplace=True)
    df = df['value'].replace('.', np.nan).astype(float)
    return df

# Fetch UK and US 10Y constant maturity yields and FEDFUNDS
us10y = fetch_fred_series("GS10")
uk10y = fetch_fred_series("IRLTLT01GBM156N")
fedfunds = fetch_fred_series("FEDFUNDS")

# Merge and calculate spread
df = pd.DataFrame({'US10Y': us10y, 'UK10Y': uk10y}).dropna()
df['SPREAD'] = df['UK10Y'] - df['US10Y']
df['FEDFUNDS'] = fedfunds.reindex(df.index).fillna(method='ffill').fillna(0)
mu = df['SPREAD'].mean()
sigma = df['SPREAD'].std()

# --- Plot 1: 0.5 Spread Chart ---
plt.figure(figsize=(12, 5))
plt.plot(df.index, df['SPREAD'], label='UK - US 10Y Spread', color='black')
plt.axhline(mu, linestyle='--', color='blue', label='Mean')
plt.axhline(mu + 0.5 * sigma, linestyle='--', color='red', label='+0.5')
plt.axhline(mu - 0.5 * sigma, linestyle='--', color='green', label='-0.5')
plt.fill_between(df.index, mu - 0.5 * sigma, mu + 0.5 * sigma, color='gray', alpha=0.2)
plt.title("UK - US 10Y Spread with 0.5 Band")
plt.ylabel('Spread (%)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# --- Plot 2: 0.75 Spread Chart ---
plt.figure(figsize=(12, 5))
plt.plot(df.index, df['SPREAD'], label='UK - US 10Y Spread', color='black')
plt.axhline(mu, linestyle='--', color='blue', label='Mean')
plt.axhline(mu + 0.75 * sigma, linestyle='--', color='red', label='+0.75')
plt.axhline(mu - 0.75 * sigma, linestyle='--', color='green', label='-0.75')
plt.fill_between(df.index, mu - 0.75 * sigma, mu + 0.75 * sigma, color='gray', alpha=0.2)
plt.title("UK - US 10Y Spread with 0.75 Band")
plt.ylabel('Spread (%)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# --- Function to compute monthly bond total return ---
def monthly_bond_return(y_prev, y_curr):
    y_prev_dec = y_prev / 100
    y_curr_dec = y_curr / 100
    T = 10.0
    carry = y_prev_dec / 12
    if y_curr_dec == 0:
        price_term = 0
    else:
        price_term = (y_prev_dec / y_curr_dec) * (1 - 1 / (1 + y_curr_dec / 2) ** (2 * (T - 1/12))) + 1 / (1 + y_curr_dec / 2) ** (2 * (T - 1/12)) - 1
    return carry + price_term

# --- Strategy Logic ---
def dislocation_strategy(df, entry_threshold):
    revert_uk = mu  # sell when reaches 0 (mean)
    revert_us = mu - 0.1 * sigma  # sell when reaches 0.10 sigma cheap
    trades = []
    in_trade = False
    trade_asset = None
    cum_ret = 1.0
    cum_ret_bench = 1.0
    prev_y = None
    entry_index = None
    entry_date = None
    monthly_strategy_ret = []
    monthly_bench_ret = []
    monthly_dates = []
    monthly_rf_list = []

    for i in range(1, len(df)):
        spread = df['SPREAD'].iloc[i]
        uk = df['UK10Y'].iloc[i]
        us = df['US10Y'].iloc[i]
        current_date = df.index[i]
        rf = df['FEDFUNDS'].iloc[i] / 100 / 12
        monthly_rf_list.append(rf)
        us_prev = df['US10Y'].iloc[i-1]
        uk_prev = df['UK10Y'].iloc[i-1]

        # Benchmark monthly always
        monthly_bench = monthly_bond_return(us_prev, us)
        monthly_bench_ret.append(monthly_bench)
        monthly_dates.append(current_date)

        # Strategy monthly and trade logic
        if in_trade:
            if trade_asset == 'UK':
                current_y = uk
            else:
                current_y = us
            monthly_ret = monthly_bond_return(prev_y, current_y)
            monthly_strategy_ret.append(monthly_ret)
            cum_ret *= (1 + monthly_ret)
            cum_ret_bench *= (1 + monthly_bench)
            prev_y = current_y
        else:
            monthly_strategy_ret.append(rf)

        # Entry check
        if not in_trade:
            if spread > mu + entry_threshold and uk > us:
                in_trade = True
                trade_asset = 'UK'
                prev_y = uk
                cum_ret = 1.0
                cum_ret_bench = 1.0
                entry_index = i
                entry_date = current_date
            elif spread < mu - entry_threshold and us > uk:
                in_trade = True
                trade_asset = 'US'
                prev_y = us
                cum_ret = 1.0
                cum_ret_bench = 1.0
                entry_index = i
                entry_date = current_date

        # Exit check
        if in_trade:
            exit_condition = False
            if trade_asset == 'UK' and spread <= revert_uk:
                exit_condition = True
            elif trade_asset == 'US' and spread >= revert_us:
                exit_condition = True
            if exit_condition:
                total_ret = (cum_ret - 1) * 100
                total_ret_bench = (cum_ret_bench - 1) * 100
                duration_months = i - entry_index
                trades.append({
                    'Entry Date': entry_date,
                    'Exit Date': current_date,
                    'Asset': trade_asset,
                    'Return (%)': total_ret,
                    'Benchmark Return (%)': total_ret_bench,
                    'Duration (Months)': duration_months
                })
                in_trade = False
                trade_asset = None
                cum_ret = 1.0
                cum_ret_bench = 1.0
                prev_y = None
                entry_index = None
                entry_date = None

    # Trades DF
    trades_df = pd.DataFrame(trades)
    trades_df['Return (%)'] = trades_df['Return (%)'].round(2)
    trades_df['Benchmark Return (%)'] = trades_df['Benchmark Return (%)'].round(2)

    # Cumulative
    strategy_cum = trades_df['Return (%)'].sum()
    bench_cum = trades_df['Benchmark Return (%)'].sum()
    difference = strategy_cum - bench_cum

    # Key ratios
    monthly_strategy_ret = np.array(monthly_strategy_ret)
    monthly_rf_list = np.array(monthly_rf_list)
    if len(monthly_strategy_ret) > 0:
        mean_monthly = np.mean(monthly_strategy_ret)
        mean_rf = np.mean(monthly_rf_list)
        std_monthly = np.std(monthly_strategy_ret)
        sharpe = (mean_monthly - mean_rf) / std_monthly * np.sqrt(12) if std_monthly > 0 else 0
        sharpe = round(sharpe, 2)

        returns = monthly_strategy_ret * 100
        sorted_returns = np.sort(returns)
        var95 = -np.percentile(sorted_returns, 5)
        var99 = -np.percentile(sorted_returns, 1)
        var95 = round(var95, 2)
        var99 = round(var99, 2)
        var95_dollar = round(var95 / 100 * 1000000, 2)
        var99_dollar = round(var99 / 100 * 1000000, 2)

        cum_wealth = np.cumprod(1 + monthly_strategy_ret)
        peak = np.maximum.accumulate(cum_wealth)
        drawdowns = (cum_wealth - peak) / peak
        max_dd = -np.min(drawdowns) * 100 if len(drawdowns) > 0 else 0
        max_dd = round(max_dd, 2)
        max_dd_dollar = round(max_dd / 100 * 1000000, 2)

        up_periods = len([r for r in monthly_strategy_ret if r > 0]) / len(monthly_strategy_ret) * 100 if len(monthly_strategy_ret) > 0 else 0
        down_periods = len([r for r in monthly_strategy_ret if r < 0]) / len(monthly_strategy_ret) * 100 if len(monthly_strategy_ret) > 0 else 0
        up_periods = round(up_periods, 2)
        down_periods = round(down_periods, 2)

        volatility = np.std(monthly_strategy_ret) * np.sqrt(12) * 100 if len(monthly_strategy_ret) > 0 else 0
        volatility = round(volatility, 2)

        ratios_df = pd.DataFrame({
            'Ratio': ['VaR 99% (%)', 'VaR 95% (%)', 'Sharpe Ratio', 'Max Drawdown (%)', '% Up Periods', '% Down Periods', 'Overall Volatility (%)'],
            'Value': [var99, var95, sharpe, max_dd, up_periods, down_periods, volatility],
            'Portfolio ($1M)': [var99_dollar, var95_dollar, '-', max_dd_dollar, '-', '-', '-']
        })
    else:
        ratios_df = pd.DataFrame()

    return trades_df, strategy_cum, bench_cum, difference, ratios_df

# Run both strategies
strategy_05_df, strategy_05_cum, bench_05_cum, diff_05, ratios_05 = dislocation_strategy(df, entry_threshold=0.5 * sigma)
strategy_075_df, strategy_075_cum, bench_075_cum, diff_075, ratios_075 = dislocation_strategy(df, entry_threshold=0.75 * sigma)

# Display for 0.5
print("Trades for 0.5 Strategy:")
display(strategy_05_df)

print("\nNet Cumulative Return for 0.5 (%):", strategy_05_cum)
print("Benchmark Cumulative Return for 0.5 (%):", bench_05_cum)
print("Difference for 0.5 (%):", diff_05)

print("\nKey Ratios for 0.5 Strategy:")
display(ratios_05)

# Display for 0.75
print("\nTrades for 0.75 Strategy:")
display(strategy_075_df)

print("\nNet Cumulative Return for 0.75 (%):", strategy_075_cum)
print("Benchmark Cumulative Return for 0.75 (%):", bench_075_cum)
print("Difference for 0.75 (%):", diff_075)

print("\nKey Ratios for 0.75 Strategy:")
display(ratios_075)
```
