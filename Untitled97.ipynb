```python
# ============================================================
# MAG7 Portfolio: Rolling Correlation, Var(Corr), Covariance
# with Equal-Weight Portfolio Returns (Ocean Theme)
# ============================================================

import pandas as pd
import numpy as np
import glob
import matplotlib.pyplot as plt

# -----------------------------
# Config
# -----------------------------
TICKERS = ["AAPL","AMZN","GOOGL","META","MSFT","NVDA","TSLA"]
WINDOW = 8   # rolling window in quarters (~2 years)
ANNUALIZE = 4

OCEAN_BG   = "#0b1e2d"
LINE_CORR  = "#17becf"
LINE_PORT  = "#ff7f0e"
WATERMARK  = "Orca's Capital Strategies"

# -----------------------------
# Load monthly returns (% change cols)
# -----------------------------
def load_monthly_returns_from_change_pct(path):
    df = pd.read_csv(path)
    df["Date"] = pd.to_datetime(df["Date"])
    r = df["Change %"].astype(str).str.replace("%","",regex=False)
    r = pd.to_numeric(r.str.replace(",","",regex=False), errors="coerce")/100.0
    s = pd.Series(r.values, index=df["Date"]).sort_index()
    return s.resample("M").last()

def monthly_to_quarterly_returns(r_monthly):
    return (1.0 + r_monthly).resample("QE").prod() - 1.0

rets_q = {}
for tkr in TICKERS:
    file = sorted(glob.glob(f"{tkr} Historical Data*.csv"))[0]
    r_m = load_monthly_returns_from_change_pct(file)
    rets_q[tkr] = monthly_to_quarterly_returns(r_m)

R = pd.DataFrame(rets_q).dropna(how="any")

# Equal-weight portfolio
R_eq = R.mean(axis=1)

# -----------------------------
# Rolling correlation stats
# -----------------------------
def rolling_avg_corr(df, window):
    out = []
    for i in range(window, len(df)+1):
        corr = df.iloc[i-window:i].corr().values
        mask = ~np.eye(len(df.columns),dtype=bool)
        out.append(corr[mask].mean())
    return pd.Series(out, index=df.index[window-1:])

def rolling_var_corr(df, window):
    out = []
    for i in range(window, len(df)+1):
        corr = df.iloc[i-window:i].corr().values
        mask = ~np.eye(len(df.columns),dtype=bool)
        out.append(corr[mask].var())
    return pd.Series(out, index=df.index[window-1:])

def rolling_avg_cov(df, window):
    out = []
    for i in range(window, len(df)+1):
        cov = df.iloc[i-window:i].cov().values
        mask = ~np.eye(len(df.columns),dtype=bool)
        out.append(cov[mask].mean())
    return pd.Series(out, index=df.index[window-1:])

roll_corr    = rolling_avg_corr(R, WINDOW)
roll_varcorr = rolling_var_corr(R, WINDOW)
roll_cov     = rolling_avg_cov(R, WINDOW)

# -----------------------------
# Rolling portfolio return (not cumulative)
# -----------------------------
roll_ret = R_eq.rolling(WINDOW).mean()

# -----------------------------
# Plot helpers
# -----------------------------
def ocean_axes(ax, title):
    ax.set_facecolor(OCEAN_BG)
    ax.figure.set_facecolor(OCEAN_BG)
    ax.set_title(title, color="white", fontsize=15, fontweight="bold", pad=18)
    ax.tick_params(colors="white")
    ax.spines["bottom"].set_color("white")
    ax.spines["left"].set_color("white")
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.grid(alpha=0.2, color="white")

def add_watermark(ax):
    ax.text(0.5, 0.05, WATERMARK, transform=ax.transAxes,
            fontsize=16, color="white", alpha=0.15,
            ha="center", va="center", rotation=30)

def plot_corr_vs_port(line, port, title, ylabel):
    fig, ax1 = plt.subplots(figsize=(12,6))
    ax1.plot(line.index, line, color=LINE_CORR, linewidth=2, label=ylabel)
    ax1.set_ylabel(ylabel, color=LINE_CORR)
    ax2 = ax1.twinx()
    ax2.plot(port.index, port, color=LINE_PORT, linewidth=1.8, label="Rolling Return")
    ax2.set_ylabel("Portfolio Return", color=LINE_PORT)
    ocean_axes(ax1, title); add_watermark(ax1)
    plt.tight_layout(); plt.show()

# -----------------------------
# Plots
# -----------------------------
plot_corr_vs_port(roll_corr, roll_ret, "Rolling Avg Correlation vs Portfolio Return", "Correlation")
plot_corr_vs_port(roll_varcorr, roll_ret, "Rolling Var(Correlation) vs Portfolio Return", "Var(Corr)")
plot_corr_vs_port(roll_cov, roll_ret, "Rolling Avg Covariance vs Portfolio Return", "Covariance")



```
