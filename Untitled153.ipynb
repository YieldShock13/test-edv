```python
# ============================================================
# FULL CONSOLIDATED PROJECT â€” DAILY 2Y/5Y/10Y BUTTERFLY (BETA-HEDGED)
# (NON-DILUTED: boundary-to-boundary, one-trade-at-a-time, NO TIME CAP)
#
# INPUTS: 3 CSV files for 2Y/5Y/10Y yields (FRED-style ok)
# OUTPUTS:
#   - boundary_reversion_summary.csv
#   - boundary_reversion_trades.csv
#   - rolling_ou_params_fly.csv
#
# CHARTS:
#   - Z-score (black bg + watermark + entry/exit lines + FINAL POINT LABELED)
#   - Rolling OU kappa (black bg + watermark + FINAL POINT LABELED)
#   - Density/Histo+KDE: holding time + trade P&L + fly_level distribution
# ============================================================

import os
import re
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.optimize import minimize
from scipy.stats import gaussian_kde

# -------------------------
# USER INPUTS (FILENAMES AS UPLOADED)
# -------------------------
Y2_FILE  = "DGS2.csv"
Y5_FILE  = "DGS5 (3).csv"
Y10_FILE = "DGS10 (13).csv"

# Rolling beta hedge (in days). 252 is ~1 trading year.
ROLL_BETA_WINDOW = 252

# Z-score window (in days). 252 is ~1 trading year.
ROLL_Z_WINDOW = 252

# Boundary-to-boundary specification (EDIT THESE TWO ONLY)
ENTRY_Z = -1.10
EXIT_Z  = -0.75

# Rolling OU window (in days). 1260 ~ 5 trading years.
ROLL_OU_WINDOW = 1260
ROLL_OU_STEP   = 5

# -------------------------
# FILE FINDER (NO /mnt IN NAMES; but robust if files landed elsewhere)
# -------------------------
def find_file(preferred_name: str) -> str:
    # 1) exact
    if os.path.exists(preferred_name):
        return preferred_name

    # 2) try common working dirs
    candidates = []
    for base in [".", "/content", "/mnt/data"]:
        p = os.path.join(base, preferred_name)
        if os.path.exists(p):
            candidates.append(p)
    if candidates:
        return candidates[0]

    # 3) fuzzy match by core token (e.g., DGS10)
    token = re.sub(r"[^A-Za-z0-9]+", "", preferred_name).lower()
    hits = []
    for base in [".", "/content", "/mnt/data"]:
        for p in glob.glob(os.path.join(base, "*.csv")):
            t = re.sub(r"[^A-Za-z0-9]+", "", os.path.basename(p)).lower()
            if token in t:
                hits.append(p)
    if hits:
        # shortest name is usually the closest match
        hits = sorted(hits, key=lambda x: (len(os.path.basename(x)), x))
        return hits[0]

    raise FileNotFoundError(f"Could not find file: {preferred_name}")

# -------------------------
# LOAD + CLEAN (FRED-style yields)
# Accepts columns like: DATE + DGS2 (or DGS5/DGS10), or Date + Value, etc.
# Missing values often '.' in FRED CSVs.
# -------------------------
def clean_numeric(x):
    if pd.isna(x):
        return np.nan
    s = str(x).strip().replace("\xa0", "")
    if s == "." or s == "":
        return np.nan
    s = s.replace(",", "")
    m = re.search(r"-?\d+(\.\d+)?", s)
    if not m:
        return np.nan
    try:
        return float(m.group(0))
    except:
        return np.nan

def load_fred_yield(csv_path: str, series_name: str) -> pd.Series:
    df = pd.read_csv(csv_path)

    # find date column
    date_col = None
    for c in df.columns:
        if c.strip().lower() in ["date", "dates", "time", "datetime", "observation_date", "date "]:
            date_col = c
            break
    if date_col is None:
        # FRED often uses "DATE"
        for c in df.columns:
            if c.strip().upper() == "DATE":
                date_col = c
                break
    if date_col is None:
        raise ValueError(f"{csv_path}: could not identify date column. Columns: {list(df.columns)}")

    # find value column
    val_col = None
    # prefer column containing series_name (e.g., DGS10)
    for c in df.columns:
        if series_name.upper() in c.upper():
            if c != date_col:
                val_col = c
                break
    if val_col is None:
        # fallback to common names
        for c in df.columns:
            if c.strip().lower() in ["value", "price", "close", "last", "rate"]:
                if c != date_col:
                    val_col = c
                    break
    if val_col is None:
        # fallback: last non-date column
        non_date_cols = [c for c in df.columns if c != date_col]
        if len(non_date_cols) == 1:
            val_col = non_date_cols[0]
        else:
            raise ValueError(f"{csv_path}: could not identify value column. Columns: {list(df.columns)}")

    df[date_col] = pd.to_datetime(df[date_col], errors="coerce")
    df[val_col]  = df[val_col].apply(clean_numeric)

    df = df.dropna(subset=[date_col, val_col]).copy()
    df = df.sort_values(date_col).drop_duplicates(subset=[date_col], keep="last")

    s = pd.Series(df[val_col].values, index=df[date_col].values, name=series_name).astype(float)

    # FRED yields are in percent; keep in percent for readability, but use bps scaling later.
    return s

# Resolve paths
Y2_PATH  = find_file(Y2_FILE)
Y5_PATH  = find_file(Y5_FILE)
Y10_PATH = find_file(Y10_FILE)

y2  = load_fred_yield(Y2_PATH,  "DGS2")
y5  = load_fred_yield(Y5_PATH,  "DGS5")
y10 = load_fred_yield(Y10_PATH, "DGS10")

# Align on common daily dates
common = y2.index.intersection(y5.index).intersection(y10.index)
y2, y5, y10 = y2.loc[common], y5.loc[common], y10.loc[common]

# -------------------------
# BETA-HEDGED BUTTERFLY (NO DV01)
# We work in daily yield CHANGES (bps) to avoid nonsense "returns" on yields.
# Regression (rolling): dy5 = a*dy2 + b*dy10 + eps
# fly_delta = dy5 - a*dy2 - b*dy10
# fly_level = cumulative sum of fly_delta (a price-like series for OU/Z)
# -------------------------
dy2  = y2.diff()
dy5  = y5.diff()
dy10 = y10.diff()

# Convert to bps for stable scaling (1% = 100 bps)
dy2_bps  = dy2 * 100.0
dy5_bps  = dy5 * 100.0
dy10_bps = dy10 * 100.0

dfd = pd.DataFrame({"dy2": dy2_bps, "dy5": dy5_bps, "dy10": dy10_bps}).dropna()

def rolling_beta_residual(dfd: pd.DataFrame, window: int) -> pd.DataFrame:
    idx = dfd.index
    a_list, b_list, resid_list = [], [], []
    out_idx = []

    x2 = dfd["dy2"].values
    x10 = dfd["dy10"].values
    y = dfd["dy5"].values

    for i in range(window, len(dfd)):
        X2 = x2[i-window:i]
        X10 = x10[i-window:i]
        Y  = y[i-window:i]

        # OLS with intercept: Y = c + a*X2 + b*X10
        # (beta-hedge uses a,b; intercept absorbed into residual)
        X = np.column_stack([np.ones(window), X2, X10])
        try:
            beta = np.linalg.lstsq(X, Y, rcond=None)[0]
            c, a, b = float(beta[0]), float(beta[1]), float(beta[2])
        except Exception:
            c, a, b = np.nan, np.nan, np.nan

        # residual at time i (current day i, using window ending at i-1)
        # IMPORTANT: no lookahead in hedge ratios:
        # the betas come from [i-window, ..., i-1], and we apply them to day i deltas.
        dy2_i, dy10_i, dy5_i = float(x2[i]), float(x10[i]), float(y[i])
        resid = dy5_i - (c + a*dy2_i + b*dy10_i)

        out_idx.append(idx[i])
        a_list.append(a)
        b_list.append(b)
        resid_list.append(resid)

    out = pd.DataFrame({"a_dy2": a_list, "b_dy10": b_list, "fly_delta_bps": resid_list}, index=pd.DatetimeIndex(out_idx))
    return out

roll_beta = rolling_beta_residual(dfd, window=ROLL_BETA_WINDOW)

# Build fly "level" series by cumulating residuals
fly_delta = roll_beta["fly_delta_bps"].copy()
fly_level = fly_delta.cumsum()
fly_level.name = "fly_level_bps"

# -------------------------
# ROLLING Z-SCORE ON FLY LEVEL (NO LOOKAHEAD)
# -------------------------
rm = fly_level.shift(1).rolling(ROLL_Z_WINDOW, min_periods=ROLL_Z_WINDOW).mean()
rs = fly_level.shift(1).rolling(ROLL_Z_WINDOW, min_periods=ROLL_Z_WINDOW).std(ddof=1)
z = (fly_level - rm) / rs
z.name = "Z_fly"

# -------------------------
# BOUNDARY-TO-BOUNDARY, NON-OVERLAPPING BACKTEST (NO TIME CAP)
# Entry: Z <= ENTRY_Z
# Exit : Z >= EXIT_Z
# Trade P&L: change in fly_level_bps from entry to exit (bps units)
# -------------------------
def run_boundary_reversion(fly_level: pd.Series, z: pd.Series, entry_z: float, exit_z: float) -> pd.DataFrame:
    df = pd.DataFrame({"fly_level_bps": fly_level, "z": z}).dropna()
    trades = []

    in_trade = False
    entry_i = None
    entry_date = None

    i = 0
    while i < len(df):
        if not in_trade:
            if df["z"].iloc[i] <= entry_z:
                in_trade = True
                entry_i = i
                entry_date = df.index[i]
        else:
            if df["z"].iloc[i] >= exit_z:
                exit_i = i
                exit_date = df.index[i]

                pnl_bps = float(df["fly_level_bps"].iloc[exit_i] - df["fly_level_bps"].iloc[entry_i])
                days_held = int((exit_date - entry_date).days)
                reverted = True

                trades.append({
                    "entry_date": entry_date,
                    "exit_date": exit_date,
                    "days_held": days_held,
                    "reverted": reverted,
                    "entry_z": float(df["z"].iloc[entry_i]),
                    "exit_z": float(df["z"].iloc[exit_i]),
                    "pnl_bps": pnl_bps,
                })

                in_trade = False
                entry_i = None
                entry_date = None
        i += 1

    # close open trade at end (not reverted)
    if in_trade and entry_i is not None:
        exit_i = len(df) - 1
        exit_date = df.index[exit_i]
        pnl_bps = float(df["fly_level_bps"].iloc[exit_i] - df["fly_level_bps"].iloc[entry_i])
        days_held = int((exit_date - entry_date).days)

        trades.append({
            "entry_date": entry_date,
            "exit_date": exit_date,
            "days_held": days_held,
            "reverted": False,
            "entry_z": float(df["z"].iloc[entry_i]),
            "exit_z": float(df["z"].iloc[exit_i]),
            "pnl_bps": pnl_bps,
        })

    return pd.DataFrame(trades)

def summarize_trades(trades: pd.DataFrame, entry_z: float, exit_z: float) -> pd.DataFrame:
    label = f"Entry Z<= {entry_z}  ->  Exit Z>= {exit_z}"
    if trades.empty:
        return pd.DataFrame([{
            "Label": label,
            "N_trades": 0,
            "Revert_rate": np.nan,
            "Median_days": np.nan,
            "Mean_days": np.nan,
            "Median_months_approx": np.nan,
            "Mean_months_approx": np.nan,
            "Mean_pnl_bps": np.nan,
            "Median_pnl_bps": np.nan,
            "HitRate_pnl": np.nan,
        }])

    pnl = trades["pnl_bps"].astype(float)
    days = trades["days_held"].astype(float)
    months_approx = days / 21.0  # trading-month approx, purely for readability

    return pd.DataFrame([{
        "Label": label,
        "N_trades": int(len(trades)),
        "Revert_rate": float(trades["reverted"].mean()),
        "Median_days": float(days.median()),
        "Mean_days": float(days.mean()),
        "Median_months_approx": float(months_approx.median()),
        "Mean_months_approx": float(months_approx.mean()),
        "Mean_pnl_bps": float(pnl.mean()),
        "Median_pnl_bps": float(pnl.median()),
        "HitRate_pnl": float((pnl > 0).mean()),
    }])

trades = run_boundary_reversion(fly_level, z, ENTRY_Z, EXIT_Z)
summary = summarize_trades(trades, ENTRY_Z, EXIT_Z)

print("\n==============================")
print("BETA-HEDGED BUTTERFLY â€” BOUNDARY MEAN REVERSION (DAILY, NON-OVERLAP, NO TIME CAP)")
print("==============================")
print(summary.to_string(index=False))

summary.to_csv("boundary_reversion_summary.csv", index=False)
trades.to_csv("boundary_reversion_trades.csv", index=False)

# -------------------------
# DENSITY / HIST + KDE PLOTS
# -------------------------
def hist_kde(data, title, xlabel, bins=40):
    x = pd.Series(data).dropna().astype(float)
    plt.figure(figsize=(11,4))
    plt.title(title)
    plt.xlabel(xlabel)
    plt.grid(True, alpha=0.2)

    if len(x) < 2 or np.isclose(x.var(), 0.0):
        plt.text(0.5, 0.5, "Not enough variation", ha="center", va="center", transform=plt.gca().transAxes)
        plt.show()
        return

    # Histogram (density)
    plt.hist(x.values, bins=bins, density=True, alpha=0.35)

    # KDE
    kde = gaussian_kde(x.values)
    grid = np.linspace(x.min(), x.max(), 400)
    plt.plot(grid, kde(grid), linewidth=2.0)
    plt.ylabel("Density")
    plt.show()

if not trades.empty:
    hist_kde(trades["days_held"], "Holding Time Density (Trades)", "Days held")
    hist_kde(trades["pnl_bps"], "Trade P&L Density (Butterfly, bps)", "P&L (bps)")

hist_kde(fly_level.dropna(), "Butterfly Level Density (fly_level_bps)", "fly_level (bps)")

# -------------------------
# BLACK-BG PLOT WITH WATERMARK + FINAL POINT LABEL
# -------------------------
def plot_black(series, title, hlines=None, final_label_fmt="{:.2f}"):
    s = pd.Series(series).dropna()
    if s.empty:
        print(f"[plot_black] No data for: {title}")
        return

    fig, ax = plt.subplots(figsize=(14,5))
    fig.patch.set_facecolor("black")
    ax.set_facecolor("black")

    ax.plot(s.index, s.values, linewidth=1.6)

    if hlines is not None:
        for yv in hlines:
            ax.axhline(yv, linewidth=1.2)

    ax.set_title(title, color="white", fontsize=14, pad=10)
    ax.tick_params(colors="white")
    for spine in ax.spines.values():
        spine.set_color("white")
    ax.grid(True, alpha=0.2)

    ax.text(0.5, 0.5, "Pearl Gray Equity and Research",
            transform=ax.transAxes, fontsize=26, color="white",
            alpha=0.08, ha="center", va="center")

    # label final point
    x_last = s.index[-1]
    y_last = float(s.iloc[-1])
    ax.scatter([x_last], [y_last], s=35)
    ax.annotate(final_label_fmt.format(y_last),
                xy=(x_last, y_last),
                xytext=(10, 10),
                textcoords="offset points",
                color="white",
                fontsize=11,
                bbox=dict(boxstyle="round,pad=0.25", fc="black", ec="white", alpha=0.85))

    plt.show()

plot_black(z, f"Daily Z-score of beta-hedged butterfly | Entry {ENTRY_Z} -> Exit {EXIT_Z}",
          hlines=[ENTRY_Z, EXIT_Z], final_label_fmt="{:.2f}")

# ============================================================
# OU MLE ON fly_level (dt=1 day)
# X_{t+1}|X_t ~ N(m, v)
# m = mu + (X_t - mu) * exp(-kappa)
# v = (sigma^2 / (2kappa)) * (1 - exp(-2kappa))
# ============================================================
def ou_neg_loglik(params, X):
    kappa, mu_, sigma = params
    if kappa <= 0 or sigma <= 0:
        return 1e50

    e = np.exp(-kappa)
    v = (sigma**2 / (2.0 * kappa)) * (1.0 - np.exp(-2.0 * kappa))
    if not np.isfinite(v) or v <= 0:
        return 1e50

    X_t = X[:-1]
    X_tp1 = X[1:]
    m = mu_ + (X_t - mu_) * e

    ll = -0.5 * np.sum(np.log(2*np.pi*v) + ((X_tp1 - m)**2) / v)
    return -ll

def fit_ou_mle(X_series: pd.Series):
    Xv = pd.Series(X_series).dropna().values.astype(float)
    if len(Xv) < 200:
        raise ValueError("Not enough daily observations to fit OU reliably (need ~200+).")

    Xt = Xv[:-1]
    Yt = Xv[1:]

    # AR(1) init: Y = a + b X
    b = np.cov(Xt, Yt, ddof=1)[0,1] / np.var(Xt, ddof=1)
    b = float(np.clip(b, -0.999, 0.999))
    a = float(np.mean(Yt) - b*np.mean(Xt))

    k0 = max(1e-6, float(-np.log(max(b, 1e-6)))) if b > 0 else 0.05
    mu0 = float(a / (1 - b)) if abs(1 - b) > 1e-6 else float(np.mean(Xv))

    eps = Yt - (a + b*Xt)
    s2 = float(np.var(eps, ddof=1))
    denom = (1 - np.exp(-2*k0))
    sig0 = np.sqrt(max(s2 * 2*k0 / max(denom, 1e-8), 1e-8))

    x0 = np.array([k0, mu0, sig0], dtype=float)
    bounds = [(1e-10, 50.0), (None, None), (1e-10, 1e6)]

    res = minimize(ou_neg_loglik, x0=x0, args=(Xv,), method="L-BFGS-B", bounds=bounds)
    if not res.success:
        raise RuntimeError(f"OU MLE failed: {res.message}")

    kappa_hat, mu_hat, sigma_hat = res.x
    return float(kappa_hat), float(mu_hat), float(sigma_hat)

kappa, mu_ou, sigma_ou = fit_ou_mle(fly_level)
half_life_days = np.log(2.0) / kappa

print("\n==============================")
print("OU FIT (fly_level_bps)")
print("==============================")
print("kappa:", round(kappa, 8))
print("mu:", round(mu_ou, 6))
print("sigma:", round(sigma_ou, 6))
print("Half-life (days):", round(half_life_days, 2))
print("Distance in z-space:", float(EXIT_Z - ENTRY_Z))

# ============================================================
# ROLLING OU (stability)
# ============================================================
def rolling_ou_params(X: pd.Series, window: int, step: int) -> pd.DataFrame:
    X = pd.Series(X).dropna()
    dates = X.index
    vals = X.values.astype(float)

    out_dates, out_k, out_mu, out_sig = [], [], [], []
    for start in range(0, len(vals) - window + 1, step):
        end = start + window
        Xw = pd.Series(vals[start:end], index=dates[start:end])
        try:
            k, m, s = fit_ou_mle(Xw)
        except Exception:
            k, m, s = np.nan, np.nan, np.nan
        out_dates.append(dates[end-1])
        out_k.append(k)
        out_mu.append(m)
        out_sig.append(s)

    return pd.DataFrame({"kappa": out_k, "mu": out_mu, "sigma": out_sig}, index=pd.DatetimeIndex(out_dates))

roll_ou = rolling_ou_params(fly_level, window=ROLL_OU_WINDOW, step=ROLL_OU_STEP)
roll_ou.to_csv("rolling_ou_params_fly.csv", index_label="Date")

plot_black(roll_ou["kappa"], f"Rolling OU kappa on fly_level_bps (window={ROLL_OU_WINDOW} days, step={ROLL_OU_STEP})",
          hlines=None, final_label_fmt="{:.6f}")

print("\nDONE.")
print("Saved: boundary_reversion_summary.csv, boundary_reversion_trades.csv, rolling_ou_params_fly.csv")
```
