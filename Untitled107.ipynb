```python
# ==== LABEL PATCH (drop-in) ====
# Adds right-edge labels for the final data points on EVERY chart,
# without changing your computations or file locations.
# Re-plots the same figures and overwrites the same PNGs with labels.

import numpy as np
import os
import matplotlib.pyplot as plt

#  helpers 
def _add_sigma_bands(ax):
    for lvl in [0, 1, 2, 3, -1, -2, -3]:
        ax.axhline(y=lvl, linestyle="--" if lvl else "-", linewidth=1, alpha=0.6)

def _add_zero_line(ax):
    ax.axhline(y=0, linestyle="--", linewidth=1, alpha=0.5)

def _wm(ax, text="orca's capital strategies"):
    ax.text(0.995, 0.02, text, transform=ax.transAxes, ha="right", va="bottom", alpha=0.5, fontsize=9)

def _label_last_series(ax, s, label):
    s = s.dropna()
    if s.empty: return
    ld = s.index.max()
    val = s.loc[ld]
    ax.annotate(f"{label}: {val:.2f}", xy=(ld, val), xytext=(10, 0),
                textcoords="offset points", va="center", fontsize=9,
                bbox=dict(boxstyle="round,pad=0.15", fc="white", ec="none", alpha=0.7))

def _label_last_df(ax, df):
    if df.empty: return
    ld = df.index.max()
    vals = df.loc[ld].dropna().sort_values()
    n = len(vals)
    nudges = np.linspace(-20, 20, n) if n > 1 else [0]
    for (name, v), dy in zip(vals.items(), nudges):
        ax.annotate(f"{name}: {v:.2f}", xy=(ld, v), xytext=(10, dy),
                    textcoords="offset points", va="center", fontsize=9,
                    bbox=dict(boxstyle="round,pad=0.15", fc="white", ec="none", alpha=0.7))

#  re-plot with labels 
# Assumes variables from your previous cell exist:
# OUT_DIR, WINDOW_WEEKS, MIN_PERIODS, z_jse_minus_spx, corr_vix_df, corr_pm_df,
# z_spreads_df, sovereign_spread, z_sovereign_spread, z_spread_yield_vs_stock, z_usdzar

# 1) Z(ALSI  SPX)
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(z_jse_minus_spx.index, z_jse_minus_spx.values, label="Z( ALSI  S&P500 )")
_add_sigma_bands(ax); _wm(ax); _label_last_series(ax, z_jse_minus_spx, "ALSIS&P500")
ax.set_title(f"Rolling Z-score: ALSI vs S&P 500 (returns)\nWindow={WINDOW_WEEKS}w")
ax.set_xlabel("Date"); ax.set_ylabel("Z-score"); ax.grid(alpha=0.3); fig.tight_layout()
fig.savefig(os.path.join(OUT_DIR, "A_Z_ALSI_minus_SPX.png"), dpi=150); plt.show()

# 2) Corr: VIX vs JSE (ALSI + sectors)
fig, ax = plt.subplots(figsize=(14,7))
for col in corr_vix_df.columns:
    ax.plot(corr_vix_df.index, corr_vix_df[col], label=col)
_add_zero_line(ax); _wm(ax); _label_last_df(ax, corr_vix_df)
ax.set_title(f"Rolling Correlation: VIX (level) vs JSE (returns)\nWindow={WINDOW_WEEKS}w")
ax.set_xlabel("Date"); ax.set_ylabel("Correlation"); ax.legend(ncol=2, frameon=False)
ax.grid(alpha=0.3); fig.tight_layout()
fig.savefig(os.path.join(OUT_DIR, "B_Corr_VIX_vs_JSE_AllSectors_MAIN.png"), dpi=150); plt.show()

for col in corr_vix_df.columns:
    fig, ax = plt.subplots(figsize=(12,6))
    ax.plot(corr_vix_df.index, corr_vix_df[col], label=col)
    _add_zero_line(ax); _wm(ax); _label_last_series(ax, corr_vix_df[col], col)
    ax.set_title(f"Rolling Corr: VIX vs {col} (returns)\nWindow={WINDOW_WEEKS}w")
    ax.set_xlabel("Date"); ax.set_ylabel("Correlation"); ax.grid(alpha=0.3); fig.tight_layout()
    fig.savefig(os.path.join(OUT_DIR, f"B_Corr_VIX_vs_{col.replace(' ','_')}.png"), dpi=150); plt.show()

# 3) Corr: Precious Metals vs JSE (ALSI + sectors)
fig, ax = plt.subplots(figsize=(14,7))
for col in corr_pm_df.columns:
    ax.plot(corr_pm_df.index, corr_pm_df[col], label=col)
_add_zero_line(ax); _wm(ax); _label_last_df(ax, corr_pm_df)
ax.set_title(f"Rolling Correlation: Precious Metals vs JSE (returns)\nWindow={WINDOW_WEEKS}w")
ax.set_xlabel("Date"); ax.set_ylabel("Correlation"); ax.legend(ncol=2, frameon=False)
ax.grid(alpha=0.3); fig.tight_layout()
fig.savefig(os.path.join(OUT_DIR, "C_Corr_PM_vs_JSE_AllSectors_MAIN.png"), dpi=150); plt.show()

for col in corr_pm_df.columns:
    fig, ax = plt.subplots(figsize=(12,6))
    ax.plot(corr_pm_df.index, corr_pm_df[col], label=col)
    _add_zero_line(ax); _wm(ax); _label_last_series(ax, corr_pm_df[col], col)
    ax.set_title(f"Rolling Corr: PM vs {col} (returns)\nWindow={WINDOW_WEEKS}w")
    ax.set_xlabel("Date"); ax.set_ylabel("Correlation"); ax.grid(alpha=0.3); fig.tight_layout()
    fig.savefig(os.path.join(OUT_DIR, f"C_Corr_PM_vs_{col.replace(' ','_')}.png"), dpi=150); plt.show()

# 4) Z-spread: z(VIX)  z(Return) for ALSI + sectors
fig, ax = plt.subplots(figsize=(14,7))
for col in z_spreads_df.columns:
    ax.plot(z_spreads_df.index, z_spreads_df[col], label=col)
_add_sigma_bands(ax); _wm(ax); _label_last_df(ax, z_spreads_df)
ax.set_title(f"VIXEquity Z-spread: z(VIX)  z(Return) (JSE)\nWindow={WINDOW_WEEKS}w")
ax.set_xlabel("Date"); ax.set_ylabel("Z-spread"); ax.legend(ncol=2, frameon=False)
ax.grid(alpha=0.3); fig.tight_layout()
fig.savefig(os.path.join(OUT_DIR, "D_Zspread_VIX_minus_Equity_MAIN.png"), dpi=150); plt.show()

for col in z_spreads_df.columns:
    fig, ax = plt.subplots(figsize=(12,6))
    ax.plot(z_spreads_df.index, z_spreads_df[col], label=col)
    _add_sigma_bands(ax); _wm(ax); _label_last_series(ax, z_spreads_df[col], col)
    ax.set_title(f"VIX{col} Z-spread (z(VIX)  z(Return))\nWindow={WINDOW_WEEKS}w")
    ax.set_xlabel("Date"); ax.set_ylabel("Z-spread"); ax.grid(alpha=0.3); fig.tight_layout()
    fig.savefig(os.path.join(OUT_DIR, f"D_Zspread_VIX_minus_{col.replace(' ','_')}.png"), dpi=150); plt.show()

# 5A) Sovereign spread (levels) + Z of spread
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(sovereign_spread.index, sovereign_spread.values, label="SA10Y  US10Y (pp)")
_add_zero_line(ax); _wm(ax); _label_last_series(ax, sovereign_spread, "SA10YUS10Y")
ax.set_title("Sovereign Spread: SA10Y  US10Y (levels)")
ax.set_xlabel("Date"); ax.set_ylabel("pp"); ax.grid(alpha=0.3); fig.tight_layout()
fig.savefig(os.path.join(OUT_DIR, "E_Spread_SA10Y_minus_US10Y_LEVELS.png"), dpi=150); plt.show()

fig, ax = plt.subplots(figsize=(12,6))
ax.plot(z_sovereign_spread.index, z_sovereign_spread.values, label="Z( SA10Y  US10Y )")
_add_sigma_bands(ax); _wm(ax); _label_last_series(ax, z_sovereign_spread, "Z(SA10YUS10Y)")
ax.set_title(f"Rolling Z: Sovereign Spread (SA10Y  US10Y)\nWindow={WINDOW_WEEKS}w")
ax.set_xlabel("Date"); ax.set_ylabel("Z-score"); ax.grid(alpha=0.3); fig.tight_layout()
fig.savefig(os.path.join(OUT_DIR, "E_Z_Spread_SA10Y_minus_US10Y.png"), dpi=150); plt.show()

# 5B) Z-spread: SA10Y(bps) vs ALSI return
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(z_spread_yield_vs_stock.index, z_spread_yield_vs_stock.values, label="z(SA10Y_bps)  z(ALSI_ret)")
_add_sigma_bands(ax); _wm(ax); _label_last_series(ax, z_spread_yield_vs_stock, "Z(SA10Y)Z(ALSI)")
ax.set_title(f"Z-spread: SA10Y (bps) vs ALSI Return\nWindow={WINDOW_WEEKS}w")
ax.set_xlabel("Date"); ax.set_ylabel("Z-spread"); ax.grid(alpha=0.3); fig.tight_layout()
fig.savefig(os.path.join(OUT_DIR, "F_Zspread_dSA10Ybps_vs_ALSI.png"), dpi=150); plt.show()

# 6) USD/ZAR rolling Z on returns
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(z_usdzar.index, z_usdzar.values, label="Z( USD/ZAR returns )")
_add_sigma_bands(ax); _wm(ax); _label_last_series(ax, z_usdzar, "USDZAR")
ax.set_title(f"Rolling Z: USD/ZAR (returns)\nWindow={WINDOW_WEEKS}w")
ax.set_xlabel("Date"); ax.set_ylabel("Z-score"); ax.grid(alpha=0.3); fig.tight_layout()
fig.savefig(os.path.join(OUT_DIR, "G_USDZAR_RollingZ.png"), dpi=150); plt.show()

print("All figures updated with right-edge labels.")

```
