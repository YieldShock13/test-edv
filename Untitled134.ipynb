```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib import font_manager as fm

import statsmodels.api as sm
from statsmodels.regression.rolling import RollingOLS
from scipy import stats

# ----------------------------
# 1) FILES (ALREADY UPLOADED)
# ----------------------------
UNRATE_FILE = "UNRATE (1).csv"
CPI_FILE    = "mcpi_revised.csv"

assert os.path.exists(UNRATE_FILE), f"File not found: {UNRATE_FILE}"
assert os.path.exists(CPI_FILE), f"File not found: {CPI_FILE}"

# ----------------------------
# 2) LOAD + CLEAN
# ----------------------------
un = pd.read_csv(UNRATE_FILE)
cpi = pd.read_csv(CPI_FILE)

un["Date"] = pd.to_datetime(un["observation_date"], errors="raise")
un["UNRATE"] = pd.to_numeric(un["UNRATE"], errors="coerce")
un = un[["Date", "UNRATE"]].dropna()

# CPI file: locate date column
date_col = next((col for col in cpi.columns if col.strip().lower() == "date"), None)
if date_col is None:
    raise ValueError(f"Could not find CPI date column. Columns: {list(cpi.columns)}")

cpi["Date"] = pd.to_datetime(cpi[date_col].astype(str).str.strip(), errors="raise")

# We need actual CPI index level if available; otherwise we reconstruct from MoM (less ideal).
# Check columns:
# - Often you'd have CPI index level. In this file we at least have monthly change + annualized MoM.
# We'll build a pseudo price index starting at 100, using monthly % change.
if "mcpi_monthly_chg" not in cpi.columns:
    raise ValueError("Expected mcpi_monthly_chg column for building CPI YoY.")
cpi["CPI_MOM"] = pd.to_numeric(cpi["mcpi_monthly_chg"], errors="coerce")
cpi = cpi[["Date", "CPI_MOM"]].dropna()

# Monthly alignment
un["Date"]  = un["Date"].dt.to_period("M").dt.to_timestamp(how="start")
cpi["Date"] = cpi["Date"].dt.to_period("M").dt.to_timestamp(how="start")

df = pd.merge(un, cpi, on="Date", how="inner").sort_values("Date").reset_index(drop=True)

# ----------------------------
# 3) BUILD INFLATION SERIES
# ----------------------------
# Construct CPI index (base=100) from monthly % changes
# CPI_MOM is assumed to be in percent units (e.g. 0.2 for +0.2%)
df["CPI_IDX"] = 100.0 * (1.0 + df["CPI_MOM"]/100.0).cumprod()

# Inflation measures
df["CPI_YOY"] = df["CPI_IDX"].pct_change(12) * 100.0          # YoY inflation (%)
df["DCPI_YOY"] = df["CPI_YOY"].diff(12)                       # change in YoY inflation over 12m (acceleration)

# drop initial NaNs
df = df.dropna().reset_index(drop=True)

# ----------------------------
# 4) BLOOMBERG-ISH STYLE
# ----------------------------
preferred_fonts = ["Arial", "Helvetica", "Roboto Condensed", "Roboto", "DejaVu Sans"]
available = {f.name for f in fm.fontManager.ttflist}
chosen = next((f for f in preferred_fonts if f in available), "DejaVu Sans")

plt.rcParams.update({
    "font.family": chosen,
    "font.size": 11,
    "axes.titlesize": 14,
    "axes.labelsize": 11,
    "xtick.labelsize": 10,
    "ytick.labelsize": 10,
})

WATERMARK = "Orca's Capital Strategies"

def add_watermark(fig):
    fig.text(0.5, 0.08, WATERMARK, ha="center", va="center",
             fontsize=22, alpha=0.12, fontweight="bold")

# ----------------------------
# 5) PHILLIPS CURVE: STATIC REGRESSION
# ----------------------------
def run_static_regression(y, x, label_y, label_x):
    X = sm.add_constant(x)
    model = sm.OLS(y, X, missing="drop").fit()
    return model

# Standard Phillips: π_t = a + b * U_t
m1 = run_static_regression(df["CPI_YOY"], df["UNRATE"], "CPI_YOY", "UNRATE")

# Accelerationist: Δπ_t = a + b * U_t
m2 = run_static_regression(df["DCPI_YOY"], df["UNRATE"], "DCPI_YOY", "UNRATE")

print("===== Phillips Curve (π_yoy vs U) =====")
print(m1.summary().tables[1])
print("\n===== Accelerationist (Δπ_yoy vs U) =====")
print(m2.summary().tables[1])

# ----------------------------
# 6) SCATTER PLOT (WITH FIT LINE)
# ----------------------------
def phillips_scatter(y, x, model, title, ylab, xlab):
    fig = plt.figure(figsize=(10.5, 6.5), dpi=140)
    ax = fig.add_subplot(111)

    ax.scatter(x, y, s=10, alpha=0.35)

    xx = np.linspace(x.min(), x.max(), 200)
    yy = model.params["const"] + model.params[x.name] * xx
    ax.plot(xx, yy, linewidth=2.2)

    ax.set_title(title)
    ax.set_xlabel(xlab)
    ax.set_ylabel(ylab)

    ax.grid(True, alpha=0.25)
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

    add_watermark(fig)
    plt.tight_layout()
    plt.show()

phillips_scatter(df["CPI_YOY"], df["UNRATE"], m1,
                 "Phillips Curve: CPI YoY Inflation vs Unemployment",
                 "CPI YoY Inflation (%)", "Unemployment Rate (%)")

phillips_scatter(df["DCPI_YOY"], df["UNRATE"], m2,
                 "Accelerationist Phillips Curve: Δ(CPI YoY) vs Unemployment",
                 "Δ(CPI YoY) over 12m (pp)", "Unemployment Rate (%)")

# ----------------------------
# 7) ROLLING REGRESSION (SLOPE THROUGH TIME)
# ----------------------------
WINDOW = 120  # 10 years monthly window (macro standard)

def rolling_beta(y, x, window=120):
    X = sm.add_constant(x)
    rols = RollingOLS(endog=y, exog=X, window=window).fit()
    return rols.params

rb1 = rolling_beta(df["CPI_YOY"], df["UNRATE"], window=WINDOW)
rb2 = rolling_beta(df["DCPI_YOY"], df["UNRATE"], window=WINDOW)

# Plot rolling beta (slope)
fig = plt.figure(figsize=(13.5, 6.2), dpi=140)
ax = fig.add_subplot(111)

ax.plot(df["Date"], rb1["UNRATE"], linewidth=2.0, label="Slope β: π_yoy ~ U (Rolling 10Y)")
ax.plot(df["Date"], rb2["UNRATE"], linewidth=2.0, linestyle="--", label="Slope β: Δπ_yoy ~ U (Rolling 10Y)")

ax.axhline(0, linewidth=1.2, alpha=0.5)
ax.set_title("Phillips Curve Slope (Rolling Regression)")
ax.set_xlabel("Date")
ax.set_ylabel("β (slope)")

ax.grid(True, alpha=0.25)
ax.spines["top"].set_visible(False)
ax.xaxis.set_major_locator(mdates.YearLocator(base=5))
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))

ax.legend(frameon=False, loc="upper right")
add_watermark(fig)
plt.tight_layout()
plt.show()

# ----------------------------
# 8) "DISCONNECT" SCORE TODAY
# ----------------------------
# Use rolling model to generate predicted inflation and compare to actual.
# We'll compute residual z-score: (actual - predicted) / rolling std(residual)

# Refit rolling predictions (π_yoy)
X = sm.add_constant(df["UNRATE"])
rols1 = RollingOLS(df["CPI_YOY"], X, window=WINDOW).fit()
df["PI_PRED"] = (rols1.params["const"] + rols1.params["UNRATE"] * df["UNRATE"])

df["RESID"] = df["CPI_YOY"] - df["PI_PRED"]
df["RESID_STD"] = df["RESID"].rolling(WINDOW).std()
df["DISCONNECT_Z"] = df["RESID"] / df["RESID_STD"]

latest = df.iloc[-1]
print("\n===== Current Phillips Disconnect =====")
print(f"Latest date: {latest['Date'].strftime('%Y-%m')}")
print(f"Unemployment: {latest['UNRATE']:.2f}%")
print(f"Actual CPI YoY: {latest['CPI_YOY']:.2f}%")
print(f"Predicted CPI YoY (based on rolling Phillips): {latest['PI_PRED']:.2f}%")
print(f"Residual (Actual - Pred): {latest['RESID']:.2f} pp")
print(f"Disconnect z-score: {latest['DISCONNECT_Z']:.2f}")

# Plot actual vs predicted and z-score
fig = plt.figure(figsize=(13.5, 6.8), dpi=140)
ax = fig.add_subplot(111)

ax.plot(df["Date"], df["CPI_YOY"], linewidth=2.0, label="Actual CPI YoY (%)")
ax.plot(df["Date"], df["PI_PRED"], linewidth=2.0, linestyle="--", label="Phillips Predicted CPI YoY (%)")

ax.set_title("CPI YoY: Actual vs Phillips Curve Implied")
ax.set_xlabel("Date")
ax.set_ylabel("Inflation (%)")

ax.grid(True, alpha=0.25)
ax.spines["top"].set_visible(False)
ax.xaxis.set_major_locator(mdates.YearLocator(base=5))
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))
ax.legend(frameon=False, loc="upper left")

# z-score on secondary axis
ax2 = ax.twinx()
ax2.plot(df["Date"], df["DISCONNECT_Z"], linewidth=1.5, alpha=0.8, label="Disconnect z-score")
ax2.axhline(0, linewidth=1.0, alpha=0.4)
ax2.set_ylabel("Z-score")

lines1, labels1 = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax.legend(lines1 + lines2, labels1 + labels2, frameon=False, loc="upper right")

add_watermark(fig)
plt.tight_layout()
plt.show()

# ============================================================
# ADD-ON SECTION (PUT BELOW YOUR EXISTING CODE)
# Regime splits + slope comparison + today's disconnect percentile
# Assumes your existing df has: Date, UNRATE, CPI_YOY, DCPI_YOY, DISCONNECT_Z
# ============================================================

import numpy as np
import pandas as pd
import statsmodels.api as sm

# ----------------------------
# 0) Sanity checks
# ----------------------------
required_cols = ["Date", "UNRATE", "CPI_YOY", "DCPI_YOY", "DISCONNECT_Z"]
missing = [c for c in required_cols if c not in df.columns]
if missing:
    raise ValueError(f"df is missing required columns for this add-on: {missing}")

df = df.sort_values("Date").reset_index(drop=True)

# ----------------------------
# 1) Regime definitions
# ----------------------------
# Use dates that match macro convention; adjust if you want.
regimes = {
    "1990-2007 (Great Moderation)": ("1990-01-01", "2007-12-31"),
    "2008-2019 (Post-GFC)":         ("2008-01-01", "2019-12-31"),
    "2020-Now (Post-COVID)":        ("2020-01-01", df["Date"].max()),
}

def fit_ols(y, x):
    X = sm.add_constant(x)
    return sm.OLS(y, X, missing="drop").fit()

rows = []
models_level = {}
models_accel = {}

for name, (start, end) in regimes.items():
    sub = df[(df["Date"] >= pd.to_datetime(start)) & (df["Date"] <= pd.to_datetime(end))].copy()
    if len(sub) < 60:
        continue

    m_level = fit_ols(sub["CPI_YOY"], sub["UNRATE"])
    m_accel = fit_ols(sub["DCPI_YOY"], sub["UNRATE"])

    models_level[name] = m_level
    models_accel[name] = m_accel

    rows.append({
        "Regime": name,
        "N": int(m_level.nobs),
        "β (π_yoy ~ U)": float(m_level.params["UNRATE"]),
        "t(β)": float(m_level.tvalues["UNRATE"]),
        "R²": float(m_level.rsquared),
        "β (Δπ_yoy ~ U)": float(m_accel.params["UNRATE"]),
        "t(β)_accel": float(m_accel.tvalues["UNRATE"]),
        "R²_accel": float(m_accel.rsquared),
    })

regime_table = pd.DataFrame(rows)
if regime_table.empty:
    raise ValueError("No regime had enough observations to estimate. Check your df date span.")

print("\n===== REGIME SPLIT RESULTS (Median CPI) =====")
display(regime_table)

# ----------------------------
# 2) Formal slope-difference test (interaction regression)
#    This tests whether slope changed in 2008+ and 2020+ relative to baseline.
# ----------------------------
# Baseline: pre-2008
df["D_2008"] = (df["Date"] >= pd.Timestamp("2008-01-01")).astype(int)
df["D_2020"] = (df["Date"] >= pd.Timestamp("2020-01-01")).astype(int)

# Interaction terms
df["U_x_D2008"] = df["UNRATE"] * df["D_2008"]
df["U_x_D2020"] = df["UNRATE"] * df["D_2020"]

# Level spec: π_yoy = a + b*U + c*D2008 + d*D2020 + e*(U*D2008) + f*(U*D2020)
X_lvl = sm.add_constant(df[["UNRATE", "D_2008", "D_2020", "U_x_D2008", "U_x_D2020"]])
m_int_lvl = sm.OLS(df["CPI_YOY"], X_lvl, missing="drop").fit()

print("\n===== SLOPE CHANGE TEST (π_yoy ~ U with interactions) =====")
print(m_int_lvl.summary().tables[1])

# Interpretation helpers:
b_base = m_int_lvl.params["UNRATE"]
b_post2008 = b_base + m_int_lvl.params["U_x_D2008"]
b_post2020 = b_base + m_int_lvl.params["U_x_D2020"]
print("\nImplied slopes:")
print(f"  Pre-2008 slope β:      {b_base:.6f}")
print(f"  Post-2008 slope β:     {b_post2008:.6f}")
print(f"  Post-2020 slope β:     {b_post2020:.6f}")
print("\nKey p-values (slope changes):")
print(f"  p(U x D_2008): {m_int_lvl.pvalues['U_x_D2008']:.6g}")
print(f"  p(U x D_2020): {m_int_lvl.pvalues['U_x_D2020']:.6g}")

# Same test for accelerationist spec
X_acc = sm.add_constant(df[["UNRATE", "D_2008", "D_2020", "U_x_D2008", "U_x_D2020"]])
m_int_acc = sm.OLS(df["DCPI_YOY"], X_acc, missing="drop").fit()

print("\n===== SLOPE CHANGE TEST (Δπ_yoy ~ U with interactions) =====")
print(m_int_acc.summary().tables[1])

b_base_a = m_int_acc.params["UNRATE"]
b_post2008_a = b_base_a + m_int_acc.params["U_x_D2008"]
b_post2020_a = b_base_a + m_int_acc.params["U_x_D2020"]
print("\nImplied slopes:")
print(f"  Pre-2008 slope β:      {b_base_a:.6f}")
print(f"  Post-2008 slope β:     {b_post2008_a:.6f}")
print(f"  Post-2020 slope β:     {b_post2020_a:.6f}")
print("\nKey p-values (slope changes):")
print(f"  p(U x D_2008): {m_int_acc.pvalues['U_x_D2008']:.6g}")
print(f"  p(U x D_2020): {m_int_acc.pvalues['U_x_D2020']:.6g}")

# ----------------------------
# 3) Today's disconnect percentile (historical distribution)
# ----------------------------
z = df["DISCONNECT_Z"].dropna().values
latest_z = float(df["DISCONNECT_Z"].dropna().iloc[-1])
pct = (z <= latest_z).mean() * 100.0

print("\n===== CURRENT DISCONNECT (Percentile) =====")
print(f"Latest date: {df['Date'].iloc[-1].strftime('%Y-%m')}")
print(f"Disconnect z-score: {latest_z:.3f}")
print(f"Percentile vs history: {pct:.1f}th")

# Optional: show a quick summary of the distribution
print("\nDisconnect z-score distribution (history):")
print(f"  mean={np.mean(z):.3f}, std={np.std(z, ddof=1):.3f}, min={np.min(z):.3f}, max={np.max(z):.3f}")
print(f"  5th={np.percentile(z,5):.3f}, 50th={np.percentile(z,50):.3f}, 95th={np.percentile(z,95):.3f}")




```
