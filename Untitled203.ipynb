```python
# ============================================================
# US 10Y YIELD vs 10Y T-NOTE FUTURES
# Time series + Z-scores + PnL + Yield PDF
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# -----------------------
# LOAD DATA
# -----------------------

price = pd.read_csv("US 10 Year T-Note Futures Historical Data (2).csv")
yield_df = pd.read_csv("United States 10-Year Bond Yield Historical Data (17).csv")

# -----------------------
# CLEAN DATA
# -----------------------

for df in [price, yield_df]:
    df["Date"] = pd.to_datetime(df["Date"])
    df.sort_values("Date", inplace=True)

price["Price"] = price["Price"].astype(str).str.replace(",","").astype(float)
yield_df["Yield"] = yield_df["Price"].astype(str).str.replace("%","").astype(float)

# -----------------------
# MERGE DATA
# -----------------------

data = pd.merge(
    price[["Date","Price"]],
    yield_df[["Date","Yield"]],
    on="Date",
    how="inner"
)

data = data.sort_values("Date").reset_index(drop=True)

# ============================================================
# POSITION / PNL PARAMETERS
# ============================================================

entry_price = 111.546875
entry_yield = 4.26
target_yield = 4.60

multiplier = 50
contracts = 1

# -----------------------
# REGRESSION PRICE ~ YIELD
# -----------------------

coef = np.polyfit(data["Yield"], data["Price"], 1)

a = coef[1]
b = coef[0]

estimated_price = a + b * target_yield

price_change = estimated_price - entry_price
pnl = price_change * multiplier * contracts

print("\n===== PnL Estimate =====")
print("Entry price:", entry_price)
print("Estimated price @",target_yield,"yield:", estimated_price)
print("Price change:", price_change)
print("PnL:", pnl)

# ============================================================
# Z-SCORES
# ============================================================

def rolling_z(series, window=252):
    mean = series.rolling(window).mean()
    std = series.rolling(window).std()
    return (series - mean) / std

data["Yield_Z"] = rolling_z(data["Yield"],252)
data["Futures_Z"] = rolling_z(data["Price"],252)

# ============================================================
# TIME SERIES CHART
# ============================================================

plt.figure(figsize=(14,6))

plt.plot(data["Date"], data["Yield"], label="US 10Y Yield", color="red")
plt.plot(data["Date"], data["Price"], label="10Y Futures Price", color="blue")

plt.title("US 10Y Yield vs Futures Price")
plt.legend()
plt.grid(True)

plt.show()

# ============================================================
# YIELD Z-SCORE
# ============================================================

plt.figure(figsize=(14,6))

plt.plot(data["Date"], data["Yield_Z"], label="Yield Z-score (252d)")
plt.axhline(2, linestyle="--")
plt.axhline(-2, linestyle="--")

plt.title("US 10Y Yield Z-score (252 day)")
plt.legend()
plt.grid(True)

plt.show()

# ============================================================
# FUTURES Z-SCORE
# ============================================================

plt.figure(figsize=(14,6))

plt.plot(data["Date"], data["Futures_Z"], label="Futures Z-score (252d)")
plt.axhline(2, linestyle="--")
plt.axhline(-2, linestyle="--")

plt.title("10Y Futures Z-score (252 day)")
plt.legend()
plt.grid(True)

plt.show()

# ============================================================
# PDF OF YIELD 400–415 BPS (LAST 3 YEARS)
# ============================================================

cutoff = data["Date"].max() - pd.DateOffset(years=3)
recent = data[data["Date"] >= cutoff].copy()

recent["Yield_bp"] = recent["Yield"] * 100

range_data = recent[(recent["Yield_bp"] >= 400) & (recent["Yield_bp"] <= 415)]

plt.figure(figsize=(14,6))

range_data["Yield_bp"].plot(
    kind="hist",
    bins=20,
    density=True,
    alpha=0.5,
    label="Histogram"
)

range_data["Yield_bp"].plot(
    kind="kde",
    linewidth=2,
    label="PDF"
)

plt.title("PDF of US 10Y Yield (400–415 bps) — Last 3 Years")
plt.xlabel("Yield (bps)")
plt.ylabel("Density")
plt.legend()
plt.grid(True)

plt.show()

print("\n===== Yield Range Stats =====")
print("Observations last 3 years:", len(recent))
print("Observations in 400–415bps:", len(range_data))
print("Probability:", len(range_data)/len(recent))
```
