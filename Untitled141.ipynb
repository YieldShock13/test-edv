```python
# ============================================================
# STXGOVJ ALL-IN-ONE:
# 1) 52-week rolling Z-score (ONLY rolling metric)
# 2) Drawdown through time (from all-time-high)
# 3) Price + running peak
# 4) "Max DD in a week"  -> worst 1-week return (separate chart)
# 5) "Max DD in 4 weeks" -> worst 4-week return (separate chart)
# 6) Max gain (max run-up from prior trough)
# Bloomberg-style charts + watermark "Pearl Gray Equity and Research"
# ============================================================

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

PATH = "/content/STXGOVJ Historical Data (5).csv"

def ensure_file_exists(path):
    if not os.path.exists(path):
        try:
            from google.colab import files
            print(f"Missing {os.path.basename(path)} in /content. Upload it now.")
            files.upload()
        except Exception:
            raise FileNotFoundError(f"File not found: {path}. Upload to /content or change PATH.")

def parse_investing_csv(path: str) -> pd.Series:
    df = pd.read_csv(path, dtype=str)
    df.columns = [c.strip().lstrip("\ufeff") for c in df.columns]

    required = {"Date", "Price"}
    if not required.issubset(set(df.columns)):
        raise ValueError(f"{os.path.basename(path)} must contain columns {required}. Found: {df.columns.tolist()}")

    # Your earlier files used YYYY/MM/DD. If this file differs, invalid rows drop out.
    df["Date"] = pd.to_datetime(df["Date"], errors="coerce", format="%Y/%m/%d")

    df["Price"] = df["Price"].astype(str).str.replace(",", "", regex=False).str.strip()
    df["Price"] = pd.to_numeric(df["Price"], errors="coerce")

    df = df.dropna(subset=["Date", "Price"]).sort_values("Date")
    return df.set_index("Date")["Price"].astype(float)

# ---------- Load ----------
ensure_file_exists(PATH)
px = parse_investing_csv(PATH).sort_index()

# ---------- 1) 52-week rolling Z-score (ROLLING ONLY HERE) ----------
WINDOW = 52
roll_mean = px.rolling(WINDOW).mean()
roll_std  = px.rolling(WINDOW).std(ddof=0)
z = (px - roll_mean) / roll_std

# ---------- 2) Drawdown through time (NOT rolling) ----------
peak = px.cummax()
drawdown = px / peak - 1.0
max_drawdown = drawdown.min()
mdd_date = drawdown.idxmin() if drawdown.notna().any() else None

# ---------- 3) Max gain (NOT rolling): max run-up from prior trough ----------
prior_trough = px.cummin()
run_up = px / prior_trough - 1.0
max_gain = run_up.max()
mg_date = run_up.idxmax() if run_up.notna().any() else None

# ---------- 4/5) Weekly and 4-week drawdowns as worst returns (NOT rolling stats) ----------
ret_1w = px.pct_change(1)
ret_4w = px.pct_change(4)

worst_1w = ret_1w.min()
worst_1w_date = ret_1w.idxmin() if ret_1w.notna().any() else None

worst_4w = ret_4w.min()
worst_4w_date = ret_4w.idxmin() if ret_4w.notna().any() else None

# ---------- Bloomberg-like styling ----------
plt.rcParams.update({
    "figure.facecolor": "black",
    "axes.facecolor": "black",
    "axes.edgecolor": "#00ff66",
    "axes.labelcolor": "#00ff66",
    "xtick.color": "#00ff66",
    "ytick.color": "#00ff66",
    "text.color": "#00ff66",
    "grid.color": "#00ff66",
    "grid.alpha": 0.18,
    "font.family": "DejaVu Sans Mono",
    "axes.titleweight": "bold",
})

def add_watermark(fig):
    fig.text(
        0.5, 0.5,
        "Orcas Capital Strategies",
        fontsize=28,
        color="#00ff66",
        alpha=0.10,
        ha="center",
        va="center",
        rotation=0
    )

# ---------- Chart A: 52-week rolling Z-score ----------
figA = plt.figure(figsize=(14, 7), dpi=150)
axA = plt.gca()
axA.plot(z.index, z.values, linewidth=1.6)
axA.axhline(0, linewidth=1.0, alpha=0.35)
axA.axhline(2, linewidth=1.0, alpha=0.25, linestyle="--")
axA.axhline(-2, linewidth=1.0, alpha=0.25, linestyle="--")
axA.set_title(f"STXGOVJ  {WINDOW}-Week Rolling Z-Score (Price)", color="#ffd24d", pad=12)
axA.set_xlabel("Date")
axA.set_ylabel("Z-score")
axA.grid(True, linestyle="-")
add_watermark(figA)

if z.notna().any():
    figA.text(0.01, 0.01, f"Latest Z: {z.dropna().iloc[-1]:.2f}", fontsize=10, alpha=0.9, ha="left", va="bottom")

plt.tight_layout()
plt.show()

# ---------- Chart B: Drawdown through time ----------
figB = plt.figure(figsize=(14, 7), dpi=150)
axB = plt.gca()
axB.plot(drawdown.index, drawdown.values * 100.0, linewidth=1.6)
axB.axhline(0, linewidth=1.0, alpha=0.35)
axB.set_title("STXGOVJ  Drawdown Through Time (from all-time-high)", color="#ffd24d", pad=12)
axB.set_xlabel("Date")
axB.set_ylabel("Drawdown (%)")
axB.grid(True, linestyle="-")
add_watermark(figB)

if mdd_date is not None:
    figB.text(
        0.01, 0.01,
        f"Max Drawdown (ATH): {max_drawdown*100:.2f}% on {mdd_date.date()}",
        fontsize=10, alpha=0.9, ha="left", va="bottom"
    )

plt.tight_layout()
plt.show()

# ---------- Chart C: Price + Running Peak ----------
figC = plt.figure(figsize=(14, 7), dpi=150)
axC = plt.gca()
axC.plot(px.index, px.values, linewidth=1.6, label="Price")
axC.plot(peak.index, peak.values, linewidth=1.0, alpha=0.6, linestyle="--", label="Running Peak")
axC.set_title("STXGOVJ  Price and Running Peak", color="#ffd24d", pad=12)
axC.set_xlabel("Date")
axC.set_ylabel("Price")
axC.grid(True, linestyle="-")
add_watermark(figC)

leg = axC.legend(frameon=False)
for t in leg.get_texts():
    t.set_color("#00ff66")

footer_parts = []
if mdd_date is not None:
    footer_parts.append(f"Max DD (ATH): {max_drawdown*100:.2f}% ({mdd_date.date()})")
if mg_date is not None:
    footer_parts.append(f"Max Gain (run-up): {max_gain*100:.2f}% ({mg_date.date()})")
if footer_parts:
    figC.text(0.01, 0.01, " | ".join(footer_parts), fontsize=10, alpha=0.9, ha="left", va="bottom")

plt.tight_layout()
plt.show()

# ---------- Chart D: 1-week returns (worst week = max 1-week drawdown) ----------
figD = plt.figure(figsize=(14, 7), dpi=150)
axD = plt.gca()
axD.plot(ret_1w.index, ret_1w.values * 100.0, linewidth=1.6)
axD.axhline(0, linewidth=1.0, alpha=0.35)
axD.set_title("STXGOVJ  1-Week Return (Worst week = max 1-week drawdown)", color="#ffd24d", pad=12)
axD.set_xlabel("Date")
axD.set_ylabel("Return (%)")
axD.grid(True, linestyle="-")
add_watermark(figD)

if worst_1w_date is not None:
    axD.scatter([worst_1w_date], [worst_1w * 100.0], s=28)
    figD.text(
        0.01, 0.01,
        f"Worst 1-week return: {worst_1w*100:.2f}% on {worst_1w_date.date()}",
        fontsize=10, alpha=0.9, ha="left", va="bottom"
    )

plt.tight_layout()
plt.show()

# ---------- Chart E: 4-week returns (worst 4-week = max 4-week drawdown) ----------
figE = plt.figure(figsize=(14, 7), dpi=150)
axE = plt.gca()
axE.plot(ret_4w.index, ret_4w.values * 100.0, linewidth=1.6)
axE.axhline(0, linewidth=1.0, alpha=0.35)
axE.set_title("STXGOVJ  4-Week Return (Worst 4-week = max 4-week drawdown)", color="#ffd24d", pad=12)
axE.set_xlabel("Date")
axE.set_ylabel("Return (%)")
axE.grid(True, linestyle="-")
add_watermark(figE)

if worst_4w_date is not None:
    axE.scatter([worst_4w_date], [worst_4w * 100.0], s=28)
    figE.text(
        0.01, 0.01,
        f"Worst 4-week return: {worst_4w*100:.2f}% on {worst_4w_date.date()}",
        fontsize=10, alpha=0.9, ha="left", va="bottom"
    )

plt.tight_layout()
plt.show()

# ---------- Output table (sanity) ----------
out = pd.DataFrame({
    "PX": px,
    f"Z_{WINDOW}w": z,
    "Peak": peak,
    "Drawdown_ATH": drawdown,
    "Ret_1w": ret_1w,
    "Ret_4w": ret_4w,
    "RunUp_from_PriorTrough": run_up,
})
display(out.tail(15))


```
