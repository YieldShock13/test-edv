```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# -----------------------
# LOAD FILES
# -----------------------

gsbd = pd.read_csv("GSBD Historical Data.csv")
hy   = pd.read_csv("(1).csv")     # your renamed HY file
dgs2 = pd.read_csv("DGS2 (1).csv")
dgs5 = pd.read_csv("DGS5 (4).csv")
dgs10= pd.read_csv("DGS10 (14).csv")
wti  = pd.read_csv("WTI_USD Historical Data (2).csv")


# -----------------------
# CLEAN INVESTING.COM DATA
# -----------------------

def clean_investing(df,name):

    df = df.copy()

    df["Date"] = pd.to_datetime(df["Date"])

    df = df[["Date","Price"]]

    df["Price"] = (
        df["Price"]
        .astype(str)
        .str.replace(",","")
        .astype(float)
    )

    df = df.rename(columns={"Price":name})

    return df.sort_values("Date")


# -----------------------
# CLEAN FRED DATA
# -----------------------

def clean_fred(df,name):

    df = df.copy()

    df["observation_date"] = pd.to_datetime(df["observation_date"])

    value_col = [c for c in df.columns if c!="observation_date"][0]

    df = df[["observation_date",value_col]]

    df.columns = ["Date",name]

    df[name] = pd.to_numeric(df[name],errors="coerce")

    return df.sort_values("Date")


# -----------------------
# APPLY CLEANING
# -----------------------

gsbd = clean_investing(gsbd,"GSBD")
hy   = clean_investing(hy,"HY_INDEX")
wti  = clean_investing(wti,"WTI")

dgs2 = clean_fred(dgs2,"UST2Y")
dgs5 = clean_fred(dgs5,"UST5Y")
dgs10= clean_fred(dgs10,"UST10Y")


# -----------------------
# MERGE DATA
# -----------------------

df = gsbd.merge(hy,on="Date",how="inner")
df = df.merge(wti,on="Date",how="inner")
df = df.merge(dgs2,on="Date",how="inner")
df = df.merge(dgs5,on="Date",how="inner")
df = df.merge(dgs10,on="Date",how="inner")

df = df.sort_values("Date").set_index("Date")


# -----------------------
# ALIGNMENT CHECK
# -----------------------

print("DATE RANGE")
print(df.index.min(),df.index.max())

print("\nMISSING VALUES")
print(df.isna().sum())


# -----------------------
# SPLIT / BAD VALUE CHECK
# -----------------------

returns = df["GSBD"].pct_change()

print("\nPotential abnormal GSBD moves (>30%)")
print(returns[np.abs(returns)>0.30])


# -----------------------
# DAILY CHANGES
# -----------------------

changes = pd.DataFrame()

changes["GSBD"] = df["GSBD"].pct_change()
changes["HY"]   = df["HY_INDEX"].pct_change()
changes["WTI"]  = df["WTI"].pct_change()

changes["UST2Y"] = df["UST2Y"].diff()
changes["UST5Y"] = df["UST5Y"].diff()
changes["UST10Y"]= df["UST10Y"].diff()

changes = changes.dropna()


# -----------------------
# FACTOR REGRESSION
# -----------------------

y = changes["GSBD"]

X = changes[["HY","WTI","UST2Y","UST5Y","UST10Y"]]

X = sm.add_constant(X)

model = sm.OLS(y,X).fit()

print(model.summary())


# -----------------------
# SHOCK ANALYSIS
# -----------------------

factor_std = changes[["HY","WTI","UST2Y","UST5Y","UST10Y"]].std()

betas = model.params[1:]

shock = pd.DataFrame({
"0.5σ":betas*factor_std*0.5,
"1σ":betas*factor_std,
"1.5σ":betas*factor_std*1.5
})

print("\nMacro Shock Impact on GSBD")
print(shock)
```
