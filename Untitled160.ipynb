```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from scipy.stats import gaussian_kde
from numpy.linalg import lstsq

# ------------------------------------------------------------
# PROFESSIONAL STYLE
# ------------------------------------------------------------

plt.style.use("default")
plt.rcParams.update({
    "font.family": "Times New Roman",
    "axes.edgecolor": "black",
    "axes.linewidth": 1.2,
    "figure.facecolor": "white",
    "axes.facecolor": "white",
    "grid.color": "grey",
    "grid.linestyle": "--",
    "grid.alpha": 0.3
})

# ------------------------------------------------------------
# 1. LOAD DATA
# ------------------------------------------------------------

nfci = pd.read_csv("NFCI (2).csv")
vix = pd.read_csv("VIXCLS (8).csv")

nfci.columns = ["Date", "NFCI"]
vix.columns = ["Date", "VIX"]

nfci["Date"] = pd.to_datetime(nfci["Date"])
vix["Date"] = pd.to_datetime(vix["Date"])

nfci = nfci.sort_values("Date").set_index("Date")
vix = vix.sort_values("Date").set_index("Date")

nfci["NFCI"] = pd.to_numeric(nfci["NFCI"], errors="coerce")
vix["VIX"] = pd.to_numeric(vix["VIX"], errors="coerce")

df = pd.concat([nfci, vix], axis=1).dropna()

print("Sample Start:", df.index.min())
print("Sample End:", df.index.max())
print("Observations:", len(df))

# ------------------------------------------------------------
# 2. Z-SCORES & PERCENTILES (52W)
# ------------------------------------------------------------

window = 52

df["NFCI_z"] = (df["NFCI"] - df["NFCI"].rolling(window).mean()) / df["NFCI"].rolling(window).std()
df["VIX_z"] = (df["VIX"] - df["VIX"].rolling(window).mean()) / df["VIX"].rolling(window).std()

df["NFCI_pct"] = df["NFCI"].rolling(window).apply(lambda x: pd.Series(x).rank(pct=True).iloc[-1])
df["VIX_pct"] = df["VIX"].rolling(window).apply(lambda x: pd.Series(x).rank(pct=True).iloc[-1])

df = df.dropna()

# ------------------------------------------------------------
# 3. REGIME GRID
# ------------------------------------------------------------

def regime(p):
    if p < 0.3:
        return "Low"
    elif p > 0.7:
        return "High"
    else:
        return "Neutral"

df["NFCI_regime"] = df["NFCI_pct"].apply(regime)
df["VIX_regime"] = df["VIX_pct"].apply(regime)

df["Regime"] = df["NFCI_regime"] + " | " + df["VIX_regime"]

print("\nRegime Counts:")
print(df["Regime"].value_counts())

# ------------------------------------------------------------
# 4. SHOCK DETECTION
# ------------------------------------------------------------

df["dNFCI"] = df["NFCI"].diff()
df["dVIX"] = df["VIX"].diff()

nfci_shock = df["dNFCI"].quantile(0.95)
vix_shock = df["dVIX"].quantile(0.95)

df["NFCI_shock"] = df["dNFCI"] > nfci_shock
df["VIX_shock"] = df["dVIX"] > vix_shock

print("\nShock Counts:")
print("NFCI:", df["NFCI_shock"].sum())
print("VIX:", df["VIX_shock"].sum())

# ------------------------------------------------------------
# 5. LAG CORRELATION
# ------------------------------------------------------------

lags = range(-8,9)
corrs = [df["dVIX"].corr(df["dNFCI"].shift(l)) for l in lags]

plt.figure(figsize=(9,4))
plt.plot(lags, corrs, color="black")
plt.axhline(0, color="black", linewidth=1)
plt.title("Lag Correlation: VIX vs NFCI")
plt.xlabel("Lag (Weeks)")
plt.ylabel("Correlation")
plt.grid(True)
plt.show()

# ------------------------------------------------------------
# 6. Z-SPREAD ANALYSIS
# ------------------------------------------------------------

df["Z_spread"] = df["NFCI_z"] - df["VIX_z"]

plt.figure(figsize=(9,4))
plt.plot(df.index, df["Z_spread"], color="black")
plt.axhline(0, color="black", linewidth=1)
plt.title("Z-Spread: NFCI_z  VIX_z")
plt.ylabel("Z-Spread")
plt.grid(True)
plt.show()

# ------------------------------------------------------------
# 7. OU HALF-LIFE ESTIMATION
# ------------------------------------------------------------

def half_life(series):
    x = series.dropna()
    y = x.diff().dropna()
    x_lag = x.shift(1).dropna()
    y = y.loc[x_lag.index]

    beta = lstsq(x_lag.values.reshape(-1,1), y.values, rcond=None)[0][0]
    halflife = -np.log(2)/beta if beta < 0 else np.nan
    return halflife

hl_nfci = half_life(df["NFCI_z"])
hl_vix = half_life(df["VIX_z"])
hl_spread = half_life(df["Z_spread"])

print("\nHalf-Life (weeks)")
print("NFCI_z:", round(hl_nfci,2))
print("VIX_z:", round(hl_vix,2))
print("Z-Spread:", round(hl_spread,2))

# ------------------------------------------------------------
# 8. DENSITY PLOTS (PROFESSIONAL)
# ------------------------------------------------------------

for col, title in zip(["NFCI_z","VIX_z","Z_spread"],
                      ["NFCI Z-Score Density",
                       "VIX Z-Score Density",
                       "Z-Spread Density"]):

    x = df[col].dropna()
    kde = gaussian_kde(x)
    xs = np.linspace(x.min(), x.max(), 200)

    plt.figure(figsize=(7,4))
    plt.plot(xs, kde(xs), color="black")
    plt.title(title)
    plt.xlabel(col)
    plt.ylabel("Density")
    plt.grid(True)
    plt.show()

# ------------------------------------------------------------
# 9. REGIME OVERLAY CHART
# ------------------------------------------------------------

plt.figure(figsize=(10,4))
plt.plot(df.index, df["NFCI"], color="black", label="NFCI")
plt.axhline(df["NFCI"].rolling(window).quantile(0.7).mean(), linestyle="--", color="grey")
plt.axhline(df["NFCI"].rolling(window).quantile(0.3).mean(), linestyle="--", color="grey")
plt.title("NFCI with Structural Percentile Bands")
plt.legend()
plt.grid(True)
plt.show()

# ------------------------------------------------------------
# 10. WALK-FORWARD REGIME PROBABILITY
# ------------------------------------------------------------

for h in [1,2,4]:
    shifted = df["VIX_regime"].shift(-h)
    prob = pd.crosstab(df["NFCI_regime"], shifted, normalize="index")
    print(f"\nProbability VIX Regime at t+{h}")
    print(prob)
```
