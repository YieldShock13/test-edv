```python
# ============================================================
# NFCI  VIX REGIME & TRANSMISSION PROJECT (WEEKLY)
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore
from statsmodels.tsa.stattools import adfuller

plt.style.use("seaborn-v0_8-white")
plt.rcParams["font.family"] = "DejaVu Serif"
plt.rcParams["axes.edgecolor"] = "black"
plt.rcParams["axes.linewidth"] = 1.0

# ============================================================
# 1. LOAD + CLEAN
# ============================================================

nfci = pd.read_csv("NFCI (2).csv")
vix = pd.read_csv("VIXCLS (8).csv")

nfci.rename(columns={nfci.columns[0]: "Date", nfci.columns[1]: "NFCI"}, inplace=True)
vix.rename(columns={vix.columns[0]: "Date", vix.columns[1]: "VIX"}, inplace=True)

nfci["Date"] = pd.to_datetime(nfci["Date"])
vix["Date"] = pd.to_datetime(vix["Date"])

nfci = nfci.sort_values("Date")
vix = vix.sort_values("Date")

nfci.set_index("Date", inplace=True)
vix.set_index("Date", inplace=True)

nfci["NFCI"] = pd.to_numeric(nfci["NFCI"], errors="coerce")
vix["VIX"] = pd.to_numeric(vix["VIX"], errors="coerce")

df = pd.concat([nfci, vix], axis=1).dropna()

print("Sample Start:", df.index.min())
print("Sample End:", df.index.max())
print("Observations:", len(df))

# ============================================================
# 2. Z-SCORES + ROLLING PERCENTILES (52W)
# ============================================================

window = 52

df["NFCI_z"] = (df["NFCI"] - df["NFCI"].rolling(window).mean()) / df["NFCI"].rolling(window).std()
df["VIX_z"] = (df["VIX"] - df["VIX"].rolling(window).mean()) / df["VIX"].rolling(window).std()

df["NFCI_pct"] = df["NFCI"].rolling(window).apply(lambda x: pd.Series(x).rank(pct=True).iloc[-1])
df["VIX_pct"] = df["VIX"].rolling(window).apply(lambda x: pd.Series(x).rank(pct=True).iloc[-1])

df = df.dropna()

# ============================================================
# 3. REGIME CLASSIFICATION (3x3 GRID)
# ============================================================

def regime_label(p):
    if p < 0.3:
        return "Low"
    elif p > 0.7:
        return "High"
    else:
        return "Neutral"

df["NFCI_regime"] = df["NFCI_pct"].apply(regime_label)
df["VIX_regime"] = df["VIX_pct"].apply(regime_label)

df["Regime_Grid"] = df["NFCI_regime"] + " | " + df["VIX_regime"]

print("\nRegime Counts:")
print(df["Regime_Grid"].value_counts())

# ============================================================
# 4. SHOCK DETECTION (95th percentile weekly change)
# ============================================================

df["dNFCI"] = df["NFCI"].diff()
df["dVIX"] = df["VIX"].diff()

nfci_shock_thresh = df["dNFCI"].quantile(0.95)
vix_shock_thresh = df["dVIX"].quantile(0.95)

df["NFCI_shock"] = df["dNFCI"] > nfci_shock_thresh
df["VIX_shock"] = df["dVIX"] > vix_shock_thresh

print("\nShock Counts:")
print("NFCI Shocks:", df["NFCI_shock"].sum())
print("VIX Shocks:", df["VIX_shock"].sum())

# ============================================================
# 5. LAG CORRELATION (Transmission)
# ============================================================

lags = range(-8, 9)
lag_corr = []

for lag in lags:
    corr = df["dVIX"].corr(df["dNFCI"].shift(lag))
    lag_corr.append(corr)

plt.figure(figsize=(8,4))
plt.plot(lags, lag_corr)
plt.axhline(0, color="black")
plt.title("Lag Correlation: VIX vs NFCI")
plt.xlabel("Lag (weeks)")
plt.ylabel("Correlation")
plt.show()

# ============================================================
# 6. IMPULSE RESPONSE STYLE AVERAGE PATH
# ============================================================

def forward_path(series, shock_col, horizon=4):
    paths = []
    shock_dates = df[df[shock_col]].index
    for date in shock_dates:
        forward = series.loc[date:].iloc[1:horizon+1].values
        if len(forward) == horizon:
            paths.append(forward)
    return np.mean(paths, axis=0)

vix_to_nfci = forward_path(df["dNFCI"], "VIX_shock", horizon=4)
nfci_to_vix = forward_path(df["dVIX"], "NFCI_shock", horizon=4)

plt.figure(figsize=(8,4))
plt.plot(range(1,5), vix_to_nfci, label="After VIX Shock")
plt.plot(range(1,5), nfci_to_vix, label="After NFCI Shock")
plt.title("Average Forward Response (1-4 Weeks)")
plt.xlabel("Weeks Ahead")
plt.ylabel("Average Change")
plt.legend()
plt.show()

# ============================================================
# 7. WALK-FORWARD REGIME TRANSITION PROBABILITY
# ============================================================

horizons = [1,2,4]

for h in horizons:
    shifted = df["VIX_regime"].shift(-h)
    prob = pd.crosstab(df["NFCI_regime"], shifted, normalize="index")
    print(f"\nProbability VIX Regime at t+{h} conditioned on NFCI Regime at t")
    print(prob)

# ============================================================
# 8. DENSITY PLOTS
# ============================================================

plt.figure(figsize=(8,4))
sns.kdeplot(df["NFCI_z"], fill=True)
plt.title("NFCI Z-Score Density")
plt.show()

plt.figure(figsize=(8,4))
sns.kdeplot(df["VIX_z"], fill=True)
plt.title("VIX Z-Score Density")
plt.show()

# ============================================================
# 9. MEAN REVERSION TEST (ADF)
# ============================================================

adf_nfci = adfuller(df["NFCI"])
adf_vix = adfuller(df["VIX"])

print("\nADF Test NFCI p-value:", adf_nfci[1])
print("ADF Test VIX p-value:", adf_vix[1])
```
