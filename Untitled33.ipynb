```python
# !pip install pandas matplotlib

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

# === Load CSV (same name, no rename needed)
df = pd.read_csv("SPY Historical Data.csv")
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
df['Price'] = df['Price'].astype(float)
df['Log_Return'] = np.log(df['Price'] / df['Price'].shift(1))
df.dropna(inplace=True)
df['Year'] = df['Date'].dt.year

# === Compute Historical VaR
hist_95, hist_99, years = [], [], []

for year, group in df.groupby('Year'):
    if len(group) < 2:
        continue
    log_returns = group['Log_Return']
    years.append(year)
    hist_95.append(100 * (1 - np.exp(log_returns.quantile(0.05))))
    hist_99.append(100 * (1 - np.exp(log_returns.quantile(0.01))))

# === Plot ===
plt.figure(figsize=(16, 6), facecolor='black')
ax = plt.gca()
ax.set_facecolor('black')

# Plot lines
plt.plot(years, hist_95, label="Historical VaR 95%", marker='o', color='deepskyblue', linewidth=2)
plt.plot(years, hist_99, label="Historical VaR 99%", marker='o', color='orange', linewidth=2)

# === Correct Final-Year Data Labels ===
# Orange (99%) goes above its line
plt.annotate(f"{hist_99[-1]:.2f}%",
             (years[-1], hist_99[-1]),
             textcoords="offset points",
             xytext=(0, 14),
             ha='center', color='orange', fontsize=9, weight='bold')

# Blue (95%) goes below its line
plt.annotate(f"{hist_95[-1]:.2f}%",
             (years[-1], hist_95[-1]),
             textcoords="offset points",
             xytext=(0, -16),
             ha='center', color='deepskyblue', fontsize=9, weight='bold')

# === Styling ===
plt.title("Historical Simulation VaR (% return)", fontsize=14, color='white', pad=20)
plt.xlabel("Year", color='white')
plt.ylabel("VaR (%)", color='white')
plt.xticks(color='white')
plt.yticks(color='white')
plt.grid(True, linestyle='dotted', color='gray', alpha=0.5)
plt.legend(facecolor='black', edgecolor='white', labelcolor='white', fontsize=10)
plt.text(0.01, 0.98, 'Pearl Gray Equity and Research',
         transform=plt.gcf().transFigure,
         fontsize=10, color='white', alpha=0.7, ha='left', va='top')
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
plt.tight_layout()
plt.show()






















```
