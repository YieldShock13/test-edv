```python
# !pip install pandas matplotlib

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

# === Hardcoded Drawdown Data ===
data = {
    "Depth": [-55.19, -47.52, -33.72, -24.5, -19.35],
    "Start": ["2007-10-10", "2000-03-27", "2020-02-20", "2022-01-04", "2018-09-21"],
    "To Bottom": [355, 637, 23, 195, 65],
    "Bottom": ["2009-03-09", "2002-10-09", "2020-03-23", "2022-10-12", "2018-12-24"],
    "To Recover": [869, 1020, 97, 294, 75],
    "End": ["2012-08-16", "2006-10-26", "2020-08-10", "2023-12-13", "2019-04-12"],
    "Total": [1224, 1657, 120, 489, 140]
}

df = pd.DataFrame(data)

# === Format for Display ===
df_display = df.copy()
df_display["Depth"] = df_display["Depth"].map(lambda x: f"{x:.2f}%")
df_display["To Bottom"] = df_display["To Bottom"].astype(str)
df_display["To Recover"] = df_display["To Recover"].astype(str)
df_display["Total"] = df_display["Total"].astype(str)

# === Plot the Table ===
plt.style.use("dark_background")
fig, ax = plt.subplots(figsize=(12, 2.5))
ax.axis('off')

table = ax.table(cellText=df_display.values,
                 colLabels=df_display.columns,
                 cellLoc='center',
                 loc='center')

# === Styling ===
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1, 1.8)

for (row, col), cell in table.get_celld().items():
    if row == 0:
        cell.set_text_props(weight='bold', color='white')
        cell.set_facecolor('#111111')
    else:
        cell.set_facecolor('#222222')

plt.title("Pearl Gray Equity and Research\nWorst Drawdowns of SPDR S&P 500 ETF", fontsize=12, pad=30)
plt.tight_layout()
plt.show()





















```
