```python
# !pip install pandas matplotlib scipy

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
from scipy.stats import norm, f_oneway
from IPython.display import display

# === 1. Load Data ===
df = pd.read_csv("SPY Historical Data.csv")
df['Date'] = pd.to_datetime(df['Date'])
df.sort_values('Date', inplace=True)
df.set_index('Date', inplace=True)
df['Price'] = df['Price'].astype(float)
df['Log_Return'] = np.log(df['Price'] / df['Price'].shift(1))
df.dropna(inplace=True)
df['Year'] = df.index.year

# === 2. Year-by-Year VaR ===
years, hist_95, hist_99, delta_95, delta_99 = [], [], [], [], []
hist_data, delta_data = [], []

for year, group in df.groupby('Year')['Log_Return']:
    if len(group) < 2:
        continue
    years.append(year)
    hist_95.append(100 * (1 - np.exp(group.quantile(0.05))))
    hist_99.append(100 * (1 - np.exp(group.quantile(0.01))))
    delta_95.append(100 * (1 - np.exp(group.mean() + norm.ppf(0.05) * group.std())))
    delta_99.append(100 * (1 - np.exp(group.mean() + norm.ppf(0.01) * group.std())))
    hist_data.append(group)
    delta_data.append(group)

# === 3. Full Sample VaR ===
hist_var_pct = {
    0.05: 100 * (1 - np.exp(df['Log_Return'].quantile(0.05))),
    0.01: 100 * (1 - np.exp(df['Log_Return'].quantile(0.01)))
}
mu, sigma = df['Log_Return'].mean(), df['Log_Return'].std()
delta_var_pct = {
    0.05: 100 * (1 - np.exp(mu + norm.ppf(0.05) * sigma)),
    0.01: 100 * (1 - np.exp(mu + norm.ppf(0.01) * sigma))
}

# === 4. Tables ===
anova_df = pd.DataFrame({
    'Model': ['Historical', 'Delta-Normal'],
    'F-Statistic': [f_oneway(*hist_data).statistic, f_oneway(*delta_data).statistic],
    'P-Value': [f_oneway(*hist_data).pvalue, f_oneway(*delta_data).pvalue]
})

summary_df = pd.DataFrame({
    'Confidence Level': ['95%', '99%'],
    'Historical VaR (%)': [hist_var_pct[0.05], hist_var_pct[0.01]],
    'Delta-Normal VaR (%)': [delta_var_pct[0.05], delta_var_pct[0.01]]
})

print("\n=== Final VaR Summary Table ===")
display(summary_df)

print("\n=== ANOVA Table ===")
display(anova_df)

# === 5. Plot: Historical VaR ===
plt.figure(figsize=(16, 6), facecolor='black')
ax = plt.gca()
ax.set_facecolor('black')

plt.plot(years, hist_95, label="Historical VaR 95%", marker='o', color='deepskyblue', linewidth=2)
plt.plot(years, hist_99, label="Historical VaR 99%", marker='o', color='orange', linewidth=2)

# Label only final year
plt.annotate(f"{hist_95[-1]:.2f}%", (years[-1], hist_95[-1]),
             textcoords="offset points", xytext=(10, 10), ha='center', color='white', fontsize=8)
plt.annotate(f"{hist_99[-1]:.2f}%", (years[-1], hist_99[-1]),
             textcoords="offset points", xytext=(10, -10), ha='center', color='white', fontsize=8)

plt.title("Historical Simulation VaR (% return)", fontsize=14, color='white', pad=20)
plt.xlabel("Year", color='white')
plt.ylabel("VaR (%)", color='white')
plt.xticks(color='white')
plt.yticks(color='white')
plt.grid(True, linestyle='dotted', color='gray', alpha=0.5)
plt.legend(facecolor='black', edgecolor='white', labelcolor='white', fontsize=10)
plt.text(0.01, 0.98, 'Pearl Gray Equity and Research',
         transform=plt.gcf().transFigure,
         fontsize=10, color='white', alpha=0.7, ha='left', va='top')
plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter())
plt.tight_layout()
plt.show()

# === 6. Plot: Delta-Normal VaR ===
plt.figure(figsize=(16, 6), facecolor='black')
ax = plt.gca()
ax.set_facecolor('black')

plt.plot(years, delta_95, label="Delta-Normal VaR 95%", marker='s', linestyle='--', color='deepskyblue', linewidth=2)
plt.plot(years, delta_99, label="Delta-Normal VaR 99%", marker='s', linestyle='--', color='orange', linewidth=2)

# Label only final year
plt.annotate(f"{delta_95[-1]:.2f}%", (years[-1], delta_95[-1]),
             textcoords="offset points", xytext=(10, 10), ha='center', color='white', fontsize=8)
plt.annotate(f"{delta_99[-1]:.2f}%", (years[-1], delta_99[-1]),
             textcoords="offset points", xytext=(10, -10), ha='center', color='white', fontsize=8)

plt.title("Delta-Normal VaR (% return)", fontsize=14, color='white', pad=20)
plt.xlabel("Year", color='white')
plt.ylabel("VaR (%)", color='white')
plt.xticks(color='white')
plt.yticks(color='white')
plt.grid(True, linestyle='dotted', color='gray', alpha=0.5)
plt.legend(facecolor='black', edgecolor='white', labelcolor='white', fontsize=10)
plt.text(0.01, 0.98, 'Pearl Gray Equity and Research',
         transform=plt.gcf().transFigure,
         fontsize=10, color='white', alpha=0.7, ha='left', va='top')
plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter())
plt.tight_layout()
plt.show()
























```
