```python
# --- If you uploaded via Colab's file uploader, files are in /content/ ---
REAL_FILE = "DFII10 (3).csv"
NOM_FILE  = "DGS10 (10).csv"
GOLD_FILE = "XAU_USD Historical Data (3).csv"

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt

# ---------- HELPERS ----------
def load_fred_csv(path, value_name_out):
    df = pd.read_csv(path)
    date_col = [c for c in df.columns if c.lower() == "date"]
    date_col = date_col[0] if date_col else df.columns[0]
    val_col = [c for c in df.columns if c != date_col][0]
    out = df[[date_col, val_col]].copy()
    out.columns = ["Date", value_name_out]
    out["Date"] = pd.to_datetime(out["Date"])
    out[value_name_out] = pd.to_numeric(
        out[value_name_out].astype(str).str.replace(",", "").str.replace("%", ""),
        errors="coerce"
    )
    return out

def load_gold_csv(path):
    df = pd.read_csv(path)
    date_col = [c for c in df.columns if c.lower() == "date"]
    date_col = date_col[0] if date_col else df.columns[0]
    out = df[[date_col, "Price"]].copy()
    out.columns = ["Date", "Gold"]
    out["Date"] = pd.to_datetime(out["Date"], errors="coerce", infer_datetime_format=True)
    out["Gold"] = pd.to_numeric(out["Gold"].astype(str).str.replace(",", ""), errors="coerce")
    return out

def add_bar_labels(ax, fmt="{:.2f}"):
    for p in ax.patches:
        h = p.get_height()
        if not np.isnan(h):
            ax.annotate(fmt.format(h),
                        (p.get_x() + p.get_width()/2.0, h),
                        ha="center", va="bottom",
                        fontsize=8, color="white",
                        xytext=(0, 3), textcoords="offset points")

# ---------- LOAD ----------
real = load_fred_csv(REAL_FILE, "RealYield")
nom  = load_fred_csv(NOM_FILE,  "NomYield")
gold = load_gold_csv(GOLD_FILE)

# ---------- MERGE & CLEAN ----------
df = real.merge(nom, on="Date", how="inner").merge(gold, on="Date", how="inner")
today = pd.Timestamp(dt.date.today())
df = df[df["Date"] <= today]
df = df.sort_values("Date").dropna()

# ---------- WEEKLY LAST OBSERVATION, LAST 12 COMPLETED WEEKS ----------
weekly_last = (
    df.set_index("Date")
      .resample("W-SUN")
      .last()
      .dropna()
)
weekly_last = weekly_last.iloc[-12:]

# ---------- WEEKLY % CHANGES ----------
weekly_changes = weekly_last.pct_change().dropna() * 100
weekly_changes.columns = ["Real Yield (%)", "Nominal Yield (%)", "Gold Price (%)"]
weekly_changes.index = weekly_changes.index.strftime("%Y-%m-%d")

# ---------- PLOT ----------
plt.style.use("dark_background")
plt.rcParams.update({
    "font.family": "DejaVu Sans",
    "font.size": 12,
    "axes.labelsize": 12,
    "axes.titlesize": 14,
    "axes.titleweight": "bold",
    "figure.titlesize": 14,
})

ax = weekly_changes.plot(kind="bar", figsize=(12, 6), alpha=0.85)
plt.title("Weekly % Changes - Real Yield, Nominal Yield, Gold (Last 12 Weeks)", pad=20)
plt.ylabel("Weekly % Change")
plt.xlabel("Week")

add_bar_labels(ax, "{:.2f}")

plt.text(0.01, 0.01, "orca's capital strategies", fontsize=10, color="gray",
         alpha=0.5, ha="left", va="bottom", transform=plt.gcf().transFigure)

plt.legend()
plt.tight_layout()
plt.show()


```
