```python
# ======================================================
# HY → IG TRANSMISSION (DAILY, 30 LAGS) — +1σ HY SHOCKS
# Units: ΔOAS in basis points (bps)
# ======================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.stats.anova import anova_lm
from statsmodels.tsa.stattools import grangercausalitytests

# ---------- File paths (update if your names differ) ----------
FILES = {
    "HY": "/content/BAMLH0A0HYM2 (9).csv",   # High Yield OAS
    "IG": "/content/BAMLC0A0CM (2).csv",     # Investment Grade OAS
}

SIGMA = 1.0        # +1σ threshold for HY shock
H = 30             # horizon & number of lags (trading days)
HAC_MAXLAGS = 5    # Newey-West bandwidth for daily data (~1 week)

# ---------- Robust CSV loader ----------
def load_oas_csv(path, name):
    df = pd.read_csv(path)
    # date column
    dcols = [c for c in df.columns if "date" in c.lower()]
    if not dcols:
        raise ValueError(f"No date column in {path}")
    dcol = dcols[0]
    df[dcol] = pd.to_datetime(df[dcol])
    df = df.sort_values(dcol).set_index(dcol)

    # pick numeric column with most non-nulls
    num_cols = []
    for c in df.columns:
        ser = pd.to_numeric(df[c].astype(str).str.replace(",", ""), errors="coerce")
        if ser.notna().sum() > 0:
            df[c] = ser
            num_cols.append(c)
    if not num_cols:
        raise ValueError(f"No numeric column in {path}")
    # choose best candidate (most observations)
    vcol = max(num_cols, key=lambda c: df[c].notna().sum())
    return df[[vcol]].rename(columns={vcol: name})

# ---------- Load & prepare daily ΔOAS in bps ----------
hy = load_oas_csv(FILES["HY"], "HY_OAS")
ig = load_oas_csv(FILES["IG"], "IG_OAS")

# If values look like %, convert to bps
def to_bps(s):
    med = s.median()
    if pd.notna(med) and 0 < med < 20:
        return s * 100.0
    return s

hy = to_bps(hy["HY_OAS"]).rename("HY_OAS").to_frame()
ig = to_bps(ig["IG_OAS"]).rename("IG_OAS").to_frame()

# Align on business days (forward-fill to avoid missing business days between sources)
hy_b = hy.asfreq("B").ffill()
ig_b = ig.asfreq("B").ffill()

# Changes in bps (daily)
dhy = hy_b.diff().rename(columns={"HY_OAS": "dHY"})
dig = ig_b.diff().rename(columns={"IG_OAS": "dIG"})

df = pd.concat([dhy, dig], axis=1).dropna().copy()

# ---------- HY +1σ shocks (widenings) ----------
thr = df["dHY"].mean() + SIGMA * df["dHY"].std()
df["HY_ShockFlag"] = (df["dHY"] > thr).astype(int)
df["HY_ShockSize"] = df["dHY"].where(df["HY_ShockFlag"] == 1, 0.0)

print(f"HY +{SIGMA}σ threshold (bps): {thr:.3f}")
print(f"Number of HY shock days: {int(df['HY_ShockFlag'].sum())}")

# ---------- A) Daily transmission path: ΔIG_{t+h} ~ β(h)*ShockSize_t ----------
def daily_path(S, y, H=30, hac_lags=5):
    rows = []
    for h in range(H+1):
        tmp = pd.DataFrame({"y": y.shift(-h), "S": S}).dropna()
        X = sm.add_constant(tmp["S"])
        # HAC for robust t/p
        hac = sm.OLS(tmp["y"], X).fit(cov_type="HAC", cov_kwds={"maxlags":hac_lags})
        # classic OLS for R2/AdjR2
        ols = sm.OLS(tmp["y"], X).fit()

        # event-only R2 (rows with S != 0)
        tmp_ev = tmp.loc[tmp["S"] != 0]
        if len(tmp_ev) >= 8:
            ols_ev = sm.OLS(tmp_ev["y"], sm.add_constant(tmp_ev["S"])).fit()
            r2_event = float(ols_ev.rsquared)
        else:
            r2_event = np.nan

        rows.append({
            "h": h,
            "beta_%perbp": float(hac.params["S"]),
            "se_hac": float(hac.bse["S"]),
            "t_hac": float(hac.tvalues["S"]),
            "p_hac": float(hac.pvalues["S"]),
            "R2_full": float(ols.rsquared),
            "AdjR2_full": float(ols.rsquared_adj),
            "R2_event": r2_event,
            "N": int(hac.nobs)
        })
    out = pd.DataFrame(rows).set_index("h")
    # Transmission horizon = last h with p<0.05
    sig = out.index[out["p_hac"] < 0.05]
    horizon = int(sig.max()) if len(sig) else 0
    return out, horizon

path, horizon = daily_path(df["HY_ShockSize"], df["dIG"], H=H, hac_lags=HAC_MAXLAGS)

print("\n=== Daily Transmission Path β(h): ΔIG_{t+h} on HY ShockSize_t ===")
print(path.to_string())
print(f"Transmission horizon (last h with p<0.05): {horizon} day(s)")

# Plot β(h) with 95% HAC CI
ci_lo = path["beta_%perbp"] - 1.96 * path["se_hac"]
ci_hi = path["beta_%perbp"] + 1.96 * path["se_hac"]
plt.figure(figsize=(10,5))
path["beta_%perbp"].plot(marker="o", title="β(h): ΔIG response to HY +1σ shock (daily, 0–30)")
plt.fill_between(path.index.values, ci_lo.values, ci_hi.values, alpha=0.2)
plt.axhline(0, ls="--", c="k")
plt.xlabel("Days after HY shock (h)")
plt.ylabel("β(h): % return in IG per 1 bp HY shock")
plt.show()

# ---------- B) Distributed-lag OLS: ΔIG_t ~ Σ_{k=0..30} ΔHY_{t-k} ----------
def build_lags(x, L):
    return pd.concat({f"x_lag{k}": x.shift(k) for k in range(L+1)}, axis=1)

Xlags = build_lags(df["dHY"], H)
df_dl = pd.concat([df["dIG"].rename("y"), Xlags], axis=1).dropna()

# Classic OLS for ANOVA, R2
rhs = " + ".join([f"x_lag{k}" for k in range(H+1)])
ols_dl = smf.ols(f"y ~ {rhs}", data=df_dl).fit()
anova = anova_lm(ols_dl, typ=2).copy()
# partial R² per lag = SS_term / (SS_term + SSE)
SSE = float(ols_dl.ssr)
anova["partial_R2"] = anova["sum_sq"] / (anova["sum_sq"] + SSE)

print("\n=== ANOVA: ΔIG_t ~ Σ ΔHY_{t-k} (k=0..30) ===")
print(anova)
print(f"Model R²: {ols_dl.rsquared:.4f} | Adj. R²: {ols_dl.rsquared_adj:.4f} | N: {int(ols_dl.nobs)}")

# HAC inference for coefs
X = sm.add_constant(df_dl[[f"x_lag{k}" for k in range(H+1)]])
hac_dl = sm.OLS(df_dl["y"], X).fit(cov_type="HAC", cov_kwds={"maxlags":HAC_MAXLAGS})
coefs = hac_dl.params.drop("const")
tvals = hac_dl.tvalues.drop("const")
pvals = hac_dl.pvalues.drop("const")
coef_table = pd.DataFrame({"coef_%perbp": coefs, "t_hac": tvals, "p_hac": pvals})
print("\n--- Distributed-lag HAC coefs (ΔIG on ΔHY lags) ---")
print(coef_table.to_string())

plt.figure(figsize=(12,5))
coefs.plot(kind="bar", title="ΔIG_t ~ ΔHY_{t−k} (k=0..30) — coefficients (% per 1 bp)")
plt.axhline(0, ls="--", c="k")
plt.ylabel("% per 1 bp")
plt.show()

# ---------- C) Granger causality (both directions), maxlag=30 ----------
# statsmodels expects array columns [y, x]
print("\n=== Granger causality: Does HY (ΔHY) Granger-cause IG (ΔIG)? ===")
_ = grangercausalitytests(df[["dIG","dHY"]].dropna(), maxlag=H, verbose=True)

print("\n=== Granger causality: Does IG (ΔIG) Granger-cause HY (ΔHY)? ===")
_ = grangercausalitytests(df[["dHY","dIG"]].dropna(), maxlag=H, verbose=True)

# ---------- D) Event study: average ΔIG path around HY shock days (±30d) ----------
def event_study_path(y, shock_flag, window=30):
    idx = y.index
    events = idx[shock_flag==1]
    paths = []
    for t in events:
        pos = idx.get_loc(t)
        if pos-window < 0 or pos+window >= len(idx):
            continue
        seg = y.iloc[pos-window:pos+window+1].reset_index(drop=True)
        seg.index = range(-window, window+1)
        paths.append(seg)
    if not paths:
        return None
    return pd.concat(paths, axis=1).mean(axis=1)

avg_path = event_study_path(df["dIG"], df["HY_ShockFlag"], window=30)
print(f"\nHY shock count used in event study: {int(df['HY_ShockFlag'].sum())}")
if avg_path is not None:
    plt.figure(figsize=(10,5))
    avg_path.plot(title="Event study: Avg ΔIG around HY +1σ shock (±30 trading days)")
    plt.axvline(0, color="r", ls="--")
    plt.xlabel("Days relative to HY shock")
    plt.ylabel("ΔIG (bps)")
    plt.show()
else:
    print("(Not enough shocks to compute event study path.)")
```
