```python
# ============================================================
# Rolling Correlation + Portfolio Risk Metrics
# Gold–Silver Pair (1σ rule | 95% winsor | 6bps | +3% cash)
# + Compare vs 100% SPX Portfolio
# ============================================================

import pandas as pd, numpy as np, matplotlib.pyplot as plt
plt.style.use("dark_background")

CASH_YIELD_ANNUAL = 0.03
CASH_MONTHLY = (1 + CASH_YIELD_ANNUAL)**(1/12) - 1
ROLL_CORR = 12
ROLL_SMOOTH = 3
LEVERAGES = [10, 20]

# ---------------- Data Loader ----------------
def load_price(fn):
    df = pd.read_csv(fn)
    df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
    df = df.dropna(subset=["Date"]).sort_values("Date")
    df["Price"] = df["Price"].astype(str).str.replace(",", "", regex=False).astype(float)
    return df.set_index("Date")["Price"]

gold   = load_price("Gold Futures Historical Data (2).csv")
silver = load_price("Silver Futures Historical Data.csv")
vix    = load_price("CBOE Volatility Index Historical Data (2).csv")
spx    = load_price("S&P 500 Historical Data (4).csv")

# ---------------- Alignment & Returns ----------------
idx = gold.index.union(silver.index).union(vix.index).union(spx.index)
gold, silver, vix, spx = [x.reindex(idx).ffill() for x in [gold, silver, vix, spx]]
gold_r, silver_r, spx_r = np.log(gold/gold.shift(1)), np.log(silver/silver.shift(1)), np.log(spx/spx.shift(1))

# ---------------- Signal Construction ----------------
vix_z = (vix - vix.rolling(12).mean()) / vix.rolling(12).std(ddof=0)
ratio = gold / silver
ratio_z = (ratio - ratio.rolling(3).mean()) / ratio.rolling(3).std(ddof=0)
sig = pd.Series(0, index=idx)
sig[(vix_z <= -1) & (ratio_z <= -1)] = -1
sig[(vix_z >=  1) & (ratio_z >=  1)] = +1
sig = sig.ffill()

# ---------------- Hedge Beta ----------------
def roll_beta(y, x, win=24):
    return y.rolling(win).cov(x)/x.rolling(win).var()

beta_gs = roll_beta(gold_r, silver_r).fillna(1).clip(0.5,1.5)
beta_sg = roll_beta(silver_r, gold_r).fillna(1).clip(0.5,1.5)

# ---------------- Spread Return ----------------
spread = pd.Series(np.nan, index=idx)
spread[sig==-1] = gold_r - beta_gs*silver_r
spread[sig==+1] = silver_r - beta_sg*gold_r

# ---------------- Vol Targeting ----------------
vol = spread.rolling(12).std(ddof=0)*np.sqrt(12)
size = (0.15/vol).clip(upper=3.0).fillna(0)

# ---------------- Strategy Return ----------------
results = {}
for L in LEVERAGES:
    ret = L*size*spread - L*0.0006          # 6bps slippage per 1x
    ret = ret.clip(upper=np.nanpercentile(ret.dropna(),95))  # winsor 95%
    ret = ret.fillna(0) + CASH_MONTHLY      # +3% cash always
    results[L] = ret.interpolate(limit_direction='both')

# ---------------- Rolling Correlation (smoothed) ----------------
for L in LEVERAGES:
    df = pd.DataFrame({"Strategy":results[L], "SPX":spx_r}).dropna()
    df["ρ"] = df["Strategy"].rolling(ROLL_CORR).corr(df["SPX"])
    df["ρ"] = df["ρ"].interpolate().rolling(ROLL_SMOOTH).mean()
    plt.figure(figsize=(12,6))
    plt.plot(df.index, df["ρ"], lw=1.8)
    plt.axhline(0, color="white", lw=1)
    plt.title(f"Rolling 12M Correlation (smoothed) — {L}× leverage | 1σ rule | +3% cash", fontsize=13)
    plt.ylabel("ρ"); plt.grid(alpha=0.3, color="white", ls="--")
    plt.tight_layout(); plt.show()

# ---------------- Risk Metric Functions ----------------
def max_drawdown(r):
    eq = np.exp(r.cumsum())
    dd = eq/eq.cummax() - 1
    return dd.min()

def value_at_risk(r, alpha=0.05):
    """VaR(95%) = loss magnitude at 5th percentile (positive value)."""
    pct = np.nanpercentile(r, 100*alpha)
    return -pct  # positive loss threshold

def portfolio_metrics(strategy_r, spx_r, w1=0.5, w2=0.5):
    df = pd.DataFrame({"Strategy":strategy_r, "SPX":spx_r}).dropna()
    port_r = w1*df["Strategy"] + w2*df["SPX"]
    ann_vol = port_r.std()*np.sqrt(12)
    max_dd  = max_drawdown(port_r)
    var95   = value_at_risk(port_r, 0.05)
    return ann_vol, max_dd, var95

# ---------------- Portfolio Risk Summary ----------------
summary = []

# Strategy + SPX (50/50)
for L in LEVERAGES:
    vol_, dd_, var_ = portfolio_metrics(results[L], spx_r)
    summary.append({
        "Portfolio":f"50/50 Strategy+SPX ({L}×)",
        "AnnVol_%":vol_*100,
        "MaxDD_%":dd_*100,
        "VaR95_%":var_*100
    })

# SPX-only (100%)
spx_vol = spx_r.std()*np.sqrt(12)
spx_dd  = max_drawdown(spx_r)
spx_var = value_at_risk(spx_r, 0.05)

summary.append({
    "Portfolio":"100% SPX Only",
    "AnnVol_%":spx_vol*100,
    "MaxDD_%":spx_dd*100,
    "VaR95_%":spx_var*100
})

summary_df = pd.DataFrame(summary)
print("\n=== Portfolio Risk Summary ===")
print(summary_df.to_string(index=False, float_format=lambda x: f"{x: .2f}"))





```
