```python
# ============================================================
# IWV vs SHYG Relative-Value Test
# Russell 3000 ETF proxy vs Short-Duration High Yield ETF proxy
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.stattools import coint, adfuller

# ----------------------------
# 1. File paths
# ----------------------------

iwv_file = "IWV Historical Data.csv"
shyg_file = "SHYG ETF Stock Price History.csv"

# ----------------------------
# 2. Cleaning function
# ----------------------------

def clean_price_file(path, ticker):
    df = pd.read_csv(path)

    df.columns = [c.strip() for c in df.columns]

    if "Date" not in df.columns or "Price" not in df.columns:
        raise ValueError(f"{ticker}: file must contain Date and Price columns.")

    df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

    df["Price"] = (
        df["Price"]
        .astype(str)
        .str.replace(",", "", regex=False)
        .str.replace("%", "", regex=False)
        .str.strip()
    )

    df["Price"] = pd.to_numeric(df["Price"], errors="coerce")

    df = df[["Date", "Price"]].dropna()
    df = df.rename(columns={"Price": ticker})
    df = df.drop_duplicates(subset="Date")
    df = df.sort_values("Date").reset_index(drop=True)

    return df

iwv = clean_price_file(iwv_file, "IWV")
shyg = clean_price_file(shyg_file, "SHYG")

# ----------------------------
# 3. Merge and sanity checks
# ----------------------------

df = pd.merge(iwv, shyg, on="Date", how="inner")
df = df.sort_values("Date").reset_index(drop=True)

print("Date range:", df["Date"].min().date(), "to", df["Date"].max().date())
print("Observations:", len(df))

print("\nMissing values:")
print(df.isna().sum())

print("\nDuplicate dates:", df["Date"].duplicated().sum())

print("\nPrice sanity:")
print(df[["IWV", "SHYG"]].describe())

# Check suspicious daily price moves
df["IWV_simple_ret"] = df["IWV"].pct_change()
df["SHYG_simple_ret"] = df["SHYG"].pct_change()

print("\nLargest absolute daily moves:")
print("IWV")
print(df.loc[df["IWV_simple_ret"].abs().nlargest(10).index, ["Date", "IWV", "IWV_simple_ret"]])

print("\nSHYG")
print(df.loc[df["SHYG_simple_ret"].abs().nlargest(10).index, ["Date", "SHYG", "SHYG_simple_ret"]])

# ----------------------------
# 4. Log returns
# ----------------------------

df["IWV_ret"] = np.log(df["IWV"]).diff()
df["SHYG_ret"] = np.log(df["SHYG"]).diff()
df = df.dropna().reset_index(drop=True)

print("\nDaily return correlation:")
print(df[["IWV_ret", "SHYG_ret"]].corr())

# ----------------------------
# 5. Rolling correlation
# ----------------------------

window = 252

df["rolling_corr_252d"] = df["IWV_ret"].rolling(window).corr(df["SHYG_ret"])

plt.figure(figsize=(12, 5))
plt.plot(df["Date"], df["rolling_corr_252d"])
plt.axhline(df["rolling_corr_252d"].mean(), linestyle="--")
plt.title("IWV vs SHYG: 252-Day Rolling Return Correlation")
plt.xlabel("Date")
plt.ylabel("Correlation")
plt.grid(True)
plt.show()

# ----------------------------
# 6. Rolling beta: SHYG returns explained by IWV returns
# ----------------------------

df["rolling_beta_252d"] = np.nan

for i in range(window, len(df)):
    y = df["SHYG_ret"].iloc[i-window:i]
    x = sm.add_constant(df["IWV_ret"].iloc[i-window:i])
    model = sm.OLS(y, x).fit()
    df.loc[i, "rolling_beta_252d"] = model.params["IWV_ret"]

plt.figure(figsize=(12, 5))
plt.plot(df["Date"], df["rolling_beta_252d"])
plt.axhline(df["rolling_beta_252d"].mean(), linestyle="--")
plt.title("Rolling Beta: SHYG Returns vs IWV Returns")
plt.xlabel("Date")
plt.ylabel("Beta")
plt.grid(True)
plt.show()

# ----------------------------
# 7. Cointegration test on log prices
# ----------------------------

df["log_IWV"] = np.log(df["IWV"])
df["log_SHYG"] = np.log(df["SHYG"])

score, pvalue, crit = coint(df["log_SHYG"], df["log_IWV"])

print("\nCointegration test: log(SHYG) ~ log(IWV)")
print("Test statistic:", score)
print("p-value:", pvalue)
print("Critical values:", crit)

# ----------------------------
# 8. Static hedge ratio and residual spread
# ----------------------------

y = df["log_SHYG"]
x = sm.add_constant(df["log_IWV"])
model = sm.OLS(y, x).fit()

df["fair_log_SHYG"] = model.predict(x)
df["spread"] = df["log_SHYG"] - df["fair_log_SHYG"]

print("\nStatic hedge regression:")
print(model.summary())

# ----------------------------
# 9. Spread z-score
# ----------------------------

z_window = 252

df["spread_mean"] = df["spread"].rolling(z_window).mean()
df["spread_std"] = df["spread"].rolling(z_window).std()
df["zscore"] = (df["spread"] - df["spread_mean"]) / df["spread_std"]

plt.figure(figsize=(12, 5))
plt.plot(df["Date"], df["zscore"])
plt.axhline(2, linestyle="--")
plt.axhline(-2, linestyle="--")
plt.axhline(0, linestyle="-")
plt.title("IWV-SHYG Relative-Value Spread Z-Score")
plt.xlabel("Date")
plt.ylabel("Z-score")
plt.grid(True)
plt.show()

# ----------------------------
# 10. ADF test on residual spread
# ----------------------------

adf_result = adfuller(df["spread"].dropna())

print("\nADF test on residual spread:")
print("ADF statistic:", adf_result[0])
print("p-value:", adf_result[1])
print("Critical values:", adf_result[4])

# ----------------------------
# 11. Half-life of mean reversion
# ----------------------------

spread = df["spread"].dropna()
spread_lag = spread.shift(1).dropna()
spread_delta = spread.diff().dropna()

half_life_df = pd.concat([spread_lag, spread_delta], axis=1).dropna()
half_life_df.columns = ["lagged_spread", "delta_spread"]

x = sm.add_constant(half_life_df["lagged_spread"])
y = half_life_df["delta_spread"]

hl_model = sm.OLS(y, x).fit()
beta = hl_model.params["lagged_spread"]

half_life = -np.log(2) / beta if beta < 0 else np.nan

print("\nHalf-life estimate:")
print("Mean-reversion beta:", beta)
print("Half-life in trading days:", half_life)

# ----------------------------
# 12. Forward return tests after extreme z-scores
# ----------------------------

for horizon in [5, 10, 21, 63]:
    df[f"IWV_fwd_{horizon}d"] = df["IWV"].shift(-horizon) / df["IWV"] - 1
    df[f"SHYG_fwd_{horizon}d"] = df["SHYG"].shift(-horizon) / df["SHYG"] - 1
    df[f"spread_fwd_change_{horizon}d"] = df["spread"].shift(-horizon) - df["spread"]

upper = df[df["zscore"] > 2]
lower = df[df["zscore"] < -2]

print("\nForward return test when spread z-score > +2")
for horizon in [5, 10, 21, 63]:
    print(
        horizon, "days |",
        "IWV:", round(upper[f"IWV_fwd_{horizon}d"].mean(), 4),
        "SHYG:", round(upper[f"SHYG_fwd_{horizon}d"].mean(), 4),
        "Spread change:", round(upper[f"spread_fwd_change_{horizon}d"].mean(), 4),
        "Obs:", upper[f"spread_fwd_change_{horizon}d"].dropna().shape[0]
    )

print("\nForward return test when spread z-score < -2")
for horizon in [5, 10, 21, 63]:
    print(
        horizon, "days |",
        "IWV:", round(lower[f"IWV_fwd_{horizon}d"].mean(), 4),
        "SHYG:", round(lower[f"SHYG_fwd_{horizon}d"].mean(), 4),
        "Spread change:", round(lower[f"spread_fwd_change_{horizon}d"].mean(), 4),
        "Obs:", lower[f"spread_fwd_change_{horizon}d"].dropna().shape[0]
    )

# ----------------------------
# 13. Residual spread plot
# ----------------------------

plt.figure(figsize=(12, 5))
plt.plot(df["Date"], df["spread"])
plt.title("Residual Spread: log(SHYG) vs log(IWV)")
plt.xlabel("Date")
plt.ylabel("Residual spread")
plt.grid(True)
plt.show()

# ----------------------------
# 14. Export cleaned dataset
# ----------------------------

df.to_csv("IWV_SHYG_cleaned_RV_test.csv", index=False)

print("\nSaved cleaned file: IWV_SHYG_cleaned_RV_test.csv")
```
