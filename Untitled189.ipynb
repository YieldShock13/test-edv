```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# =========================
# CONFIG
# =========================
FILE_USDTHB = "USD_THB Historical Data (1).csv"
FILE_GSBD   = "GSBD Historical Data.csv"

NOTIONAL_USDTHB = 2000.0          # USD
NOTIONAL_GSBD   = 1800.0      # USD

ROLL_DAILY  = 252
ROLL_WEEKLY = 52
ROLL_MONTH  = 12

VAR_LEVEL = 0.95  # 95% VaR

# =========================
# LOAD + CLEAN (Investing.com format)
# =========================
def load_investing_price_csv(path, out_name):
    df = pd.read_csv(path).copy()
    df.columns = [c.strip() for c in df.columns]

    if "Date" not in df.columns or "Price" not in df.columns:
        raise ValueError(f"{path}: expected columns including 'Date' and 'Price'. Found: {df.columns.tolist()}")

    # Investing.com sometimes uses YYYY/MM/DD, sometimes MM/DD/YYYY; let pandas infer.
    df["Date"] = pd.to_datetime(df["Date"], infer_datetime_format=True, errors="coerce")

    # numeric parse for "Price" (commas)
    df["Price"] = (
        df["Price"]
        .astype(str)
        .str.replace(",", "", regex=False)
        .str.replace("%", "", regex=False)
    )
    df["Price"] = pd.to_numeric(df["Price"], errors="coerce")

    df = df.dropna(subset=["Date", "Price"]).sort_values("Date")
    df = df[["Date", "Price"]].rename(columns={"Price": out_name}).set_index("Date")
    return df

usdthb = load_investing_price_csv(FILE_USDTHB, "USDTHB")
gsbd   = load_investing_price_csv(FILE_GSBD, "GSBD")

# =========================
# ALIGN + RETURNS
# =========================
prices = usdthb.join(gsbd, how="inner").sort_index()

# simple returns
rets_d = prices.pct_change().dropna()

# =========================
# RESAMPLE TO WEEKLY + MONTHLY RETURNS
# (use last observation of period)
# =========================
prices_w = prices.resample("W-FRI").last().dropna()
prices_m = prices.resample("M").last().dropna()

rets_w = prices_w.pct_change().dropna()
rets_m = prices_m.pct_change().dropna()

# =========================
# ROLLING CORRELATIONS
# =========================
rollcorr_d = rets_d["USDTHB"].rolling(ROLL_DAILY).corr(rets_d["GSBD"])
rollcorr_w = rets_w["USDTHB"].rolling(ROLL_WEEKLY).corr(rets_w["GSBD"])
rollcorr_m = rets_m["USDTHB"].rolling(ROLL_MONTH).corr(rets_m["GSBD"])

# =========================
# BACKTEST: PnL, RETURNS, DRAWDOWN
# - Define PnL in USD as notional * return
# - Portfolio return measured vs initial total capital (NOTIONAL_USDTHB + NOTIONAL_GSBD)
# =========================
def backtest_pnl(returns_df, n1, n2, name1="USDTHB", name2="GSBD"):
    r1 = returns_df["USDTHB"]
    r2 = returns_df["GSBD"]

    pnl1 = n1 * r1
    pnl2 = n2 * r2
    pnlP = pnl1 + pnl2

    init_cap = n1 + n2
    ret1 = pnl1 / n1
    ret2 = pnl2 / n2
    retP = pnlP / init_cap

    eq1 = (1 + ret1).cumprod()
    eq2 = (1 + ret2).cumprod()
    eqP = (1 + retP).cumprod()

    dd1 = eq1 / eq1.cummax() - 1
    dd2 = eq2 / eq2.cummax() - 1
    ddP = eqP / eqP.cummax() - 1

    out = pd.DataFrame({
        "pnl_"+name1: pnl1,
        "pnl_"+name2: pnl2,
        "pnl_port": pnlP,
        "ret_"+name1: ret1,
        "ret_"+name2: ret2,
        "ret_port": retP,
        "eq_"+name1: eq1,
        "eq_"+name2: eq2,
        "eq_port": eqP,
        "dd_"+name1: dd1,
        "dd_"+name2: dd2,
        "dd_port": ddP,
    }, index=returns_df.index)

    return out

bt_d = backtest_pnl(rets_d, NOTIONAL_USDTHB, NOTIONAL_GSBD)

# =========================
# RISK METRICS: daily vol, VaR 95%, max drawdown
# =========================
def risk_metrics(bt_df):
    # daily vol of returns (not annualized)
    vol = bt_df[["ret_USDTHB", "ret_GSBD", "ret_port"]].std()

    # historical VaR 95% (1-day): VaR is positive number representing loss threshold
    # VaR_95 = -quantile(1-0.95) = -5th percentile
    var = -bt_df[["ret_USDTHB", "ret_GSBD", "ret_port"]].quantile(1 - VAR_LEVEL)

    # max drawdown
    mdd = bt_df[["dd_USDTHB", "dd_GSBD", "dd_port"]].min()

    return vol, var, mdd

vol_d, var_d, mdd_d = risk_metrics(bt_d)

# =========================
# RISK CONTRIBUTION (2-asset) on DAILY returns
# Use return covariance and weights = dollar notionals / total capital
# RC_i to variance: w_i * (Σ w)_i / (w' Σ w)
# =========================
def risk_contribution_two_asset(returns_df, n1, n2):
    r = returns_df[["USDTHB", "GSBD"]].dropna()
    cov = r.cov().values
    w = np.array([n1, n2], dtype=float)
    w = w / w.sum()  # portfolio return weights

    port_var = float(w.T @ cov @ w)
    m = cov @ w  # marginal contribution vector (Σ w)

    rc = w * m / port_var
    return {
        "weights": w,
        "cov": cov,
        "port_var": port_var,
        "rc": rc
    }

rc_out = risk_contribution_two_asset(rets_d, NOTIONAL_USDTHB, NOTIONAL_GSBD)

# =========================
# EQUAL RISK CONTRIBUTION: solve for USDTHB notional so RC_USDTHB = RC_GSBD
# Holding GSBD notional fixed, vary USDTHB notional.
# Work in "PnL-variance" space using raw dollar weights (not normalized).
# Condition RC1 = RC2 where RC_i = w_i * (Σ w)_i, with w = [n1, n2].
# Solve: n1*(n1*s11 + n2*s12) = n2*(n1*s12 + n2*s22)
# => n1^2*s11 = n2^2*s22  (cross terms cancel)
# => n1 = n2 * sqrt(s22/s11)
# This holds for 2-asset case with that RC definition.
# =========================
def equal_rc_notional_usdthb(returns_df, n_gsbd_fixed):
    r = returns_df[["USDTHB", "GSBD"]].dropna()
    cov = r.cov()
    s11 = float(cov.loc["USDTHB", "USDTHB"])
    s22 = float(cov.loc["GSBD", "GSBD"])
    if s11 <= 0 or s22 <= 0:
        raise ValueError("Non-positive variances encountered; cannot compute equal risk notionals.")
    n1 = n_gsbd_fixed * np.sqrt(s22 / s11)
    return n1, cov

eq_notional_usdthb, cov_d = equal_rc_notional_usdthb(rets_d, NOTIONAL_GSBD)

# =========================
# OUTPUTS
# =========================
print("\n====================")
print("DATE ALIGNMENT")
print(prices.index.min(), "to", prices.index.max())
print("Daily obs:", len(rets_d), "| Weekly obs:", len(rets_w), "| Monthly obs:", len(rets_m))

print("\n====================")
print("ROLLING CORRELATION (last available)")
print("Daily 252d:", float(rollcorr_d.dropna().iloc[-1]) if rollcorr_d.dropna().shape[0] else np.nan)
print("Weekly 52w:", float(rollcorr_w.dropna().iloc[-1]) if rollcorr_w.dropna().shape[0] else np.nan)
print("Monthly 12m:", float(rollcorr_m.dropna().iloc[-1]) if rollcorr_m.dropna().shape[0] else np.nan)

print("\n====================")
print("RISK METRICS (DAILY)")
risk_table = pd.DataFrame({
    "daily_vol": vol_d,
    "VaR_95_1d": var_d,
    "max_drawdown": mdd_d
})
print(risk_table)

print("\n====================")
print("RISK CONTRIBUTION (DAILY, return-weights)")
w = rc_out["weights"]
rc = rc_out["rc"]
rc_table = pd.DataFrame({
    "weight": [w[0], w[1]],
    "risk_contribution_share": [rc[0], rc[1]],
}, index=["USDTHB", "GSBD"])
print(rc_table)

print("\n====================")
print("EQUAL RISK CONTRIBUTION NOTIONAL (DAILY covariance)")
print(f"Current notionals: USDTHB={NOTIONAL_USDTHB:,.2f} , GSBD={NOTIONAL_GSBD:,.2f}")
print(f"USDTHB notional for ~50/50 variance contribution (GSBD fixed): {eq_notional_usdthb:,.2f} USD")

# =========================
# PLOTS
# =========================

# Rolling correlations
plt.figure(figsize=(10,4))
plt.plot(rollcorr_d.index, rollcorr_d.values)
plt.title("Rolling Corr: USDTHB vs GSBD (Daily, 252d)")
plt.axhline(0)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10,4))
plt.plot(rollcorr_w.index, rollcorr_w.values)
plt.title("Rolling Corr: USDTHB vs GSBD (Weekly, 52w)")
plt.axhline(0)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10,4))
plt.plot(rollcorr_m.index, rollcorr_m.values)
plt.title("Rolling Corr: USDTHB vs GSBD (Monthly, 12m)")
plt.axhline(0)
plt.tight_layout()
plt.show()

# Equity curves (normalized)
plt.figure(figsize=(10,4))
plt.plot(bt_d.index, bt_d["eq_USDTHB"], label="USDTHB")
plt.plot(bt_d.index, bt_d["eq_GSBD"], label="GSBD")
plt.plot(bt_d.index, bt_d["eq_port"], label="Portfolio")
plt.title("Equity Curves (Daily, normalized)")
plt.legend()
plt.tight_layout()
plt.show()

# Drawdowns
plt.figure(figsize=(10,4))
plt.plot(bt_d.index, bt_d["dd_USDTHB"], label="USDTHB")
plt.plot(bt_d.index, bt_d["dd_GSBD"], label="GSBD")
plt.plot(bt_d.index, bt_d["dd_port"], label="Portfolio")
plt.title("Drawdowns (Daily)")
plt.legend()
plt.tight_layout()
plt.show()
```
