```python
# ============================================================
# STAT-ARB (WEEKLY)  SHORT XAGUSD, LONG (beta-adjusted) XAUUSD
#
# YOU ASKED FOR: SILVER RELATIVE TO GOLD (NOT GOLD RELATIVE TO SILVER)
# Spread (silver relative to gold):
#   s_t = log(XAG_t) - beta_t * log(XAU_t)
# where beta_t is estimated from: r_xag = a + beta * r_xau
#
# TRADE RULE (ONE-SIDED):
# - When spread is "1 sigma RICH" (z >= +1.0): short silver + long beta*gold (1.00x)
# - If already in at 1.0 and gets richer: add +0.50x at z >= +1.5
# - Add +0.25x at z >= +2.0
# - EXIT ALL when z <= +0.25 (still rich but mean-reverted enough)
#
# Capital: $12,000
# Costs:
#  - Fixed: $12 round-trip equivalent -> model as $6 per trade EVENT (add/exit)
#  - Slippage: 12 bps per trade EVENT on traded gross notional
#
# Files: weekly Investing.com-style "Historical Data" CSVs
# ============================================================

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller, kpss, coint
from statsmodels.stats.diagnostic import acorr_ljungbox
from statsmodels.tsa.vector_ar.vecm import coint_johansen
from statsmodels.stats.stattools import jarque_bera
from statsmodels.stats.diagnostic import het_arch

# ----------------------------
# 1) FILES
# ----------------------------
XAU_FILE = "XAU_USD Historical Data (7).csv"
XAG_FILE = "XAG_USD Historical Data (2).csv"

# ----------------------------
# 2) CAPITAL + COSTS
# ----------------------------
CAPITAL_USD = 12_000.0

EVENT_FIXED_COST_USD = 6.0     # per event (each add/exit)
SLIPPAGE_BPS = 12.0
SLIPPAGE = SLIPPAGE_BPS / 10_000.0

MAX_GROSS_LEV = 1.0            # gross notional <= equity * MAX_GROSS_LEV

# ----------------------------
# 3) PARAMS (weekly)
# ----------------------------
ROLL_BETA_W = 104
ZSCORE_W    = 104

USE_KALMAN_BETA = True
USE_ROLLING_BETA_FALLBACK = True

# Ladder thresholds: "rich" = positive z (silver rich vs gold)
Z_L1   =  1.0
Z_L2   =  1.5
Z_L3   =  2.0
Z_EXIT =  0.25

ALLOC_L1 = 1.00
ALLOC_L2 = 0.50
ALLOC_L3 = 0.25

REQUIRE_VALID_Z = True

# ----------------------------
# 4) IO helpers
# ----------------------------
def _find_file(fname: str) -> str:
    candidates = [
        fname,
        os.path.join("/content", fname),
        os.path.join("/mnt/data", fname),
    ]
    for p in candidates:
        if os.path.exists(p):
            return p
    raise FileNotFoundError(f"Could not find {fname}. Upload it to Colab (/content) and retry.")

def _to_float_series(s: pd.Series) -> pd.Series:
    return pd.to_numeric(s.astype(str).str.replace(",", "").str.strip(), errors="coerce")

def load_investing_hist_csv(path: str) -> pd.DataFrame:
    df = pd.read_csv(path).copy()
    df["Date"] = pd.to_datetime(df["Date"], errors="coerce", infer_datetime_format=True)
    df = df.dropna(subset=["Date"])
    df["Price"] = _to_float_series(df["Price"])
    df = df.sort_values("Date").set_index("Date")
    return df[["Price"]].dropna()

def logret(px: pd.Series) -> pd.Series:
    return np.log(px).diff()

# ----------------------------
# 5) Stats helpers
# ----------------------------
def adf_test(x: pd.Series, name: str):
    x = x.dropna()
    stat, p, lags, nobs, crit, icbest = adfuller(x, autolag="AIC")
    return {"series": name, "ADF_stat": float(stat), "pvalue": float(p), "lags": int(lags), "nobs": int(nobs),
            "crit_1%": float(crit.get("1%")), "crit_5%": float(crit.get("5%")), "crit_10%": float(crit.get("10%"))}

def kpss_test(x: pd.Series, name: str, regression="c"):
    x = x.dropna()
    stat, p, lags, crit = kpss(x, regression=regression, nlags="auto")
    return {"series": name, "KPSS_stat": float(stat), "pvalue": float(p), "lags": int(lags), "nobs": int(len(x)),
            "crit_10%": float(crit.get("10%")), "crit_5%": float(crit.get("5%")), "crit_2.5%": float(crit.get("2.5%")),
            "crit_1%": float(crit.get("1%"))}

def z_from_roll(spread: pd.Series, window: int):
    m = spread.rolling(window).mean()
    s = spread.rolling(window).std(ddof=0)
    return (spread - m) / s

def ljung_box_p(x: pd.Series, lags=10):
    x = x.dropna()
    if len(x) < lags + 5:
        return np.nan
    out = acorr_ljungbox(x, lags=[lags], return_df=True)
    return float(out["lb_pvalue"].iloc[0])

def rolling_acf1(x: pd.Series, window: int):
    out = pd.Series(index=x.index, dtype=float)
    for i in range(window, len(x) + 1):
        w = x.iloc[i-window:i].dropna()
        if len(w) < window * 0.9:
            continue
        out.iloc[i-1] = float(w.autocorr(lag=1))
    return out

def shannon_entropy(x: pd.Series, bins=30):
    v = x.dropna().values
    if len(v) < 100:
        return np.nan
    hist, _ = np.histogram(v, bins=bins, density=True)
    p = hist[hist > 0]
    return float(-np.sum(p * np.log(p)))

def rolling_entropy(x: pd.Series, window: int, bins=30):
    out = pd.Series(index=x.index, dtype=float)
    for i in range(window, len(x) + 1):
        w = x.iloc[i-window:i]
        out.iloc[i-1] = shannon_entropy(w, bins=bins)
    return out

def ou_half_life(spread: pd.Series):
    x = spread.dropna()
    if len(x) < 60:
        return {"half_life_weeks": np.nan, "b": np.nan, "R2": np.nan}
    x_lag = x.shift(1).dropna()
    dx = x.diff().dropna()
    idx = dx.index.intersection(x_lag.index)
    X = sm.add_constant(x_lag.loc[idx].values)
    fit = sm.OLS(dx.loc[idx].values, X).fit()
    b = float(fit.params[1])
    r2 = float(fit.rsquared)
    if b >= 0:
        return {"half_life_weeks": float(np.inf), "b": b, "R2": r2}
    hl = float(-np.log(2) / b)
    return {"half_life_weeks": hl, "b": b, "R2": r2}

def rolling_ols_beta(y: pd.Series, x: pd.Series, window: int):
    idx = y.dropna().index.intersection(x.dropna().index)
    y = y.loc[idx]; x = x.loc[idx]
    out = pd.Series(index=idx, dtype=float)
    for i in range(window, len(idx) + 1):
        widx = idx[i-window:i]
        yy = y.loc[widx].values
        xx = x.loc[widx].values
        X = sm.add_constant(xx)
        b = sm.OLS(yy, X).fit().params[1]
        out.loc[widx[-1]] = float(b)
    return out

def kalman_beta(y: pd.Series, x: pd.Series, delta=1e-5, R=1e-3):
    """
    1D Kalman filter for time-varying beta in: y_t = beta_t * x_t + e_t
    """
    idx = y.dropna().index.intersection(x.dropna().index)
    y = y.loc[idx].astype(float)
    x = x.loc[idx].astype(float)

    beta = 0.0
    P = 1.0
    Q = delta / (1 - delta)

    out = []
    for yt, xt in zip(y.values, x.values):
        P = P + Q
        if np.isfinite(xt) and abs(xt) > 1e-12 and np.isfinite(yt):
            H = xt
            K = P * H / (H * P * H + R)
            beta = beta + K * (yt - H * beta)
            P = (1 - K * H) * P
        out.append(beta)

    return pd.Series(out, index=idx, name="beta_kalman")

def perf_stats(rets_log: pd.Series):
    rets = rets_log.dropna()
    if len(rets) < 50:
        return {}
    ann = 52
    cagr = float(np.exp(rets.mean() * ann) - 1.0)
    vol  = float(rets.std(ddof=0) * np.sqrt(ann))
    sharpe = float((rets.mean() * ann) / (rets.std(ddof=0) * np.sqrt(ann) + 1e-12))
    eq = np.exp(rets.cumsum())
    dd = eq / eq.cummax() - 1.0
    maxdd = float(dd.min())
    return {"CAGR": cagr, "Vol": vol, "Sharpe": sharpe, "MaxDD": maxdd, "N_weeks": int(len(rets))}

# ----------------------------
# 6) LOAD + ALIGN
# ----------------------------
xau = load_investing_hist_csv(_find_file(XAU_FILE))
xag = load_investing_hist_csv(_find_file(XAG_FILE))

px = pd.DataFrame({"XAU": xau["Price"], "XAG": xag["Price"]}).dropna()

data = pd.DataFrame({
    "log_xau": np.log(px["XAU"]),
    "log_xag": np.log(px["XAG"]),
    "r_xau": logret(px["XAU"]),
    "r_xag": logret(px["XAG"]),
}).dropna()

idx = data.index
print("Aligned weekly sample:", idx.min().date(), "to", idx.max().date(), "N=", len(data))

print("Return sanity (weekly):")
print("  XAU |r|>20% freq:", float((data["r_xau"].abs() > 0.20).mean()))
print("  XAG |r|>20% freq:", float((data["r_xag"].abs() > 0.20).mean()))

# ----------------------------
# 7) BETA (SILVER ~ beta * GOLD)
# r_xag = a + beta * r_xau
# ----------------------------
Y = data["r_xag"].values
X = sm.add_constant(data["r_xau"].values)
beta_static = float(sm.OLS(Y, X).fit().params[1])

beta_used = pd.Series(beta_static, index=idx)

beta_kf = kalman_beta(data["r_xag"], data["r_xau"]) if USE_KALMAN_BETA else None
beta_roll = rolling_ols_beta(data["r_xag"], data["r_xau"], window=ROLL_BETA_W) if USE_ROLLING_BETA_FALLBACK else None

if USE_KALMAN_BETA and beta_kf is not None:
    beta_used = beta_kf.reindex(idx).ffill()
elif beta_roll is not None:
    beta_used = beta_roll.reindex(idx).ffill()

print("Beta static (r_xag ~ beta*r_xau):", beta_static)
print("Beta used last:", float(beta_used.dropna().iloc[-1]))

beta_abs = beta_used.abs().clip(lower=1e-6)

# ----------------------------
# 8) SPREAD (SILVER RELATIVE TO GOLD) + SIGMA
# s_t = log(XAG_t) - beta_t * log(XAU_t)
# z = (s - rollmean)/rollstd
# ----------------------------
spread = data["log_xag"] - beta_used * data["log_xau"]
z = z_from_roll(spread, window=ZSCORE_W)

# ----------------------------
# 9) LADDER UNITS (silver rich => positive z)
# Enter/add on z >= +1 / +1.5 / +2
# Exit all on z <= +0.25
# ----------------------------
units = pd.Series(0.0, index=idx)

state_L1 = False
state_L2 = False
state_L3 = False

for t in idx:
    zt = z.loc[t]

    if REQUIRE_VALID_Z and (not np.isfinite(zt)):
        units.loc[t] = units.shift(1).loc[t] if t != idx[0] else 0.0
        continue

    # exit all when it mean-reverts down to +0.25 rich or less
    if np.isfinite(zt) and (zt <= Z_EXIT):
        state_L1 = state_L2 = state_L3 = False

    # ladder adds when it is rich
    if np.isfinite(zt) and (zt >= Z_L1):
        state_L1 = True
    if np.isfinite(zt) and (zt >= Z_L2):
        state_L2 = True
    if np.isfinite(zt) and (zt >= Z_L3):
        state_L3 = True

    u = 0.0
    if state_L1: u += ALLOC_L1
    if state_L2: u += ALLOC_L2
    if state_L3: u += ALLOC_L3

    units.loc[t] = u

# ----------------------------
# 10) WEIGHTS (SHORT SILVER, LONG beta*GOLD), gross-capped
# Base scale so that at units=1, gross = MAX_GROSS_LEV:
# gross = |w_silver| + |w_gold| = scale*(1 + |beta|)
# => scale = MAX_GROSS_LEV/(1+|beta|)
# Then apply units multiplier.
# ----------------------------
base_scale = (MAX_GROSS_LEV / (1.0 + beta_abs)).clip(upper=MAX_GROSS_LEV)

w_silver = units * base_scale              # SHORT this amount
w_gold   = units * base_scale * beta_abs   # LONG this amount (beta-adjusted)

# Portfolio weekly log-return approximation:
# short silver => -w_silver*r_xag
# long gold    => +w_gold*r_xau
r_gross = (-w_silver * data["r_xag"]) + (w_gold * data["r_xau"])
r_gross = r_gross.fillna(0.0)

eq_gross = np.exp(r_gross.cumsum()) * CAPITAL_USD

# ----------------------------
# 11) COSTS (fixed + slippage per event on traded gross notional)
# Event = any change in units (add or exit)
# ----------------------------
d_units = units.diff().fillna(units)
events = (d_units.abs() > 1e-12).astype(float)

# gross fraction = |w_silver| + |w_gold|
gross_frac = (w_silver.abs() + w_gold.abs()).fillna(0.0)

# traded gross fraction approx from delta units:
traded_gross_frac = (d_units.abs() * base_scale * (1.0 + beta_abs)).fillna(0.0)

eq_prev = eq_gross.shift(1).fillna(CAPITAL_USD)
traded_gross_notional = traded_gross_frac * eq_prev

fixed_cost_usd = events * EVENT_FIXED_COST_USD
slip_cost_usd  = traded_gross_notional * SLIPPAGE
total_cost_usd = fixed_cost_usd + slip_cost_usd

cost_ret = (total_cost_usd / eq_prev.replace(0, np.nan)).fillna(0.0)

r_net = r_gross - cost_ret
eq_net = np.exp(r_net.cumsum()) * CAPITAL_USD

# ----------------------------
# 12) FULL TEST SUITE (GLOBAL + ROLLING)
# ----------------------------
print("\n================ FULL TEST SUITE (GLOBAL) ================")

print("\n--- Unit root tests: log levels ---")
print(adf_test(data["log_xau"], "log_xau"))
print(kpss_test(data["log_xau"], "log_xau"))
print(adf_test(data["log_xag"], "log_xag"))
print(kpss_test(data["log_xag"], "log_xag"))

print("\n--- Engle-Granger: log_xag ~ log_xau ---")
# EG cointegration test on log levels:
# coint(y, x) tests y - b*x residual stationarity
tstat, pval, crit = coint(data["log_xag"], data["log_xau"])
print({"pair": "log_xag~log_xau", "EG_tstat": float(tstat), "pvalue": float(pval), "crit_5%": float(crit[1])})

# Also show OLS slope / R2 for context
X_eg = sm.add_constant(data["log_xau"].values)
fit_eg = sm.OLS(data["log_xag"].values, X_eg).fit()
print("EG slope:", float(fit_eg.params[1]), "R2:", float(fit_eg.rsquared))

eg_resid = pd.Series(fit_eg.resid, index=idx, name="eg_resid")
print("ADF on EG residual:", adf_test(eg_resid, "eg_resid"))
print("KPSS on EG residual:", kpss_test(eg_resid, "eg_resid"))

print("\n--- Johansen cointegration (2-var, log levels) ---")
try:
    j = coint_johansen(data[["log_xau", "log_xag"]].dropna(), det_order=0, k_ar_diff=1)
    print({
        "trace_stat_r0": float(j.lr1[0]), "trace_crit5_r0": float(j.cvt[0, 1]),
        "trace_stat_r1": float(j.lr1[1]), "trace_crit5_r1": float(j.cvt[1, 1]),
        "eig_stat_r0": float(j.lr2[0]), "eig_crit5_r0": float(j.cvm[0, 1]),
        "eig_stat_r1": float(j.lr2[1]), "eig_crit5_r1": float(j.cvm[1, 1]),
    })
except Exception as e:
    print("Johansen failed:", str(e))

print("\n--- Spread (constructed: log_xag - beta*log_xau) diagnostics ---")
print("ADF(spread):", adf_test(spread, "spread"))
print("KPSS(spread):", kpss_test(spread, "spread"))
print("ACF1(spread):", float(spread.dropna().autocorr(lag=1)))
print("Ljung-Box p (10) spread:", ljung_box_p(spread, lags=10))
print("OU half-life:", ou_half_life(spread))
print("Shannon entropy (global):", shannon_entropy(spread))

print("\n--- Spread return diagnostics (portfolio gross returns) ---")
jb_stat, jb_p, skew, kurt = jarque_bera(r_gross.dropna())
print("Jarque-Bera:", float(jb_stat), "p:", float(jb_p), "skew:", float(skew), "kurtosis:", float(kurt))
print("Ljung-Box p (10) r_gross:", ljung_box_p(r_gross, lags=10))
arch_stat, arch_p, _, _ = het_arch(r_gross.dropna(), nlags=10)
print("ARCH(10) stat:", float(arch_stat), "p:", float(arch_p))
print("mean weekly:", float(r_gross.mean()), "std weekly:", float(r_gross.std(ddof=0)))

# Rolling stability
ROLL_STAB_W = 156
roll_adf_p = spread.rolling(ROLL_STAB_W).apply(lambda x: adfuller(pd.Series(x).dropna(), autolag="AIC")[1] if pd.Series(x).dropna().shape[0] >= 60 else np.nan, raw=False)
roll_kpss_p = spread.rolling(ROLL_STAB_W).apply(lambda x: kpss(pd.Series(x).dropna(), regression="c", nlags="auto")[1] if pd.Series(x).dropna().shape[0] >= 60 else np.nan, raw=False)
roll_acf1 = rolling_acf1(spread, ROLL_STAB_W)
roll_vol = spread.rolling(ROLL_STAB_W).std(ddof=0)
roll_ent = rolling_entropy(spread, ROLL_STAB_W, bins=30)

print("\n================ ROLLING STABILITY (WINDOW=156w) ================")
print("Latest rolling ADF p:", float(roll_adf_p.dropna().iloc[-1]) if roll_adf_p.dropna().size else np.nan)
print("Latest rolling KPSS p:", float(roll_kpss_p.dropna().iloc[-1]) if roll_kpss_p.dropna().size else np.nan)
print("Latest rolling ACF1  :", float(roll_acf1.dropna().iloc[-1]) if roll_acf1.dropna().size else np.nan)
print("Latest rolling vol   :", float(roll_vol.dropna().iloc[-1]) if roll_vol.dropna().size else np.nan)
print("Latest rolling ent   :", float(roll_ent.dropna().iloc[-1]) if roll_ent.dropna().size else np.nan)

# ----------------------------
# 13) BACKTEST RESULTS
# ----------------------------
print("\n================ BACKTEST RESULTS (SIGMA LADDER) ================")
print("Gross:", perf_stats(r_gross))
print("Net  :", perf_stats(r_net))
print("Trade events (adds/exits counted):", int(events.sum()))

# ----------------------------
# 14) P&L LEDGER (EVENT-BASED)
# Each event row: date, z, units_before/after, traded notional, costs, equity_before/after, leg contribs
# ----------------------------
silver_leg = (-w_silver * data["r_xag"]).fillna(0.0)
gold_leg   = ( w_gold   * data["r_xau"]).fillna(0.0)

ledger = pd.DataFrame(index=idx)
ledger["z"] = z
ledger["beta_used"] = beta_used
ledger["units_before"] = units.shift(1).fillna(0.0)
ledger["units_after"]  = units
ledger["delta_units"]  = d_units
ledger["w_silver_short"] = w_silver
ledger["w_gold_long"] = w_gold
ledger["equity_before"] = eq_prev
ledger["equity_after_gross"] = eq_gross
ledger["equity_after_net"]   = eq_net
ledger["traded_gross_notional_usd"] = traded_gross_notional
ledger["fixed_cost_usd"] = fixed_cost_usd
ledger["slip_cost_usd"]  = slip_cost_usd
ledger["total_cost_usd"] = total_cost_usd
ledger["cost_ret"] = cost_ret
ledger["silver_leg_logret"] = silver_leg
ledger["gold_leg_logret"]   = gold_leg
ledger["gross_logret"] = r_gross
ledger["net_logret"]   = r_net

ledger_events = ledger.loc[events.astype(bool)].copy()
ledger_events.to_csv("xag_xau_sigma_ladder_LEDGER.csv")

# ----------------------------
# 15) FULL TIMESERIES OUTPUT
# ----------------------------
out = pd.DataFrame({
    "XAU": px["XAU"].reindex(idx),
    "XAG": px["XAG"].reindex(idx),
    "log_xau": data["log_xau"],
    "log_xag": data["log_xag"],
    "r_xau": data["r_xau"],
    "r_xag": data["r_xag"],
    "beta_used": beta_used,
    "spread_silver_rel_gold": spread,
    "z": z,
    "units": units,
    "w_silver_short": w_silver,
    "w_gold_long": w_gold,
    "gross_frac": gross_frac,
    "traded_gross_notional_usd": traded_gross_notional,
    "fixed_cost_usd": fixed_cost_usd,
    "slip_cost_usd": slip_cost_usd,
    "total_cost_usd": total_cost_usd,
    "r_gross": r_gross,
    "r_net": r_net,
    "eq_gross_usd": eq_gross,
    "eq_net_usd": eq_net,
    "roll_adf_p_156w": roll_adf_p,
    "roll_kpss_p_156w": roll_kpss_p,
    "roll_acf1_156w": roll_acf1,
    "roll_vol_156w": roll_vol,
    "roll_entropy_156w": roll_ent,
}).dropna(how="all")

out.to_csv("xag_short_xau_long_sigma_ladder_timeseries.csv")

print("\nSaved: xag_short_xau_long_sigma_ladder_timeseries.csv")
print("Saved: xag_xau_sigma_ladder_LEDGER.csv")

# ----------------------------
# 16) PLOTS
# ----------------------------
plt.figure()
plt.plot(px.index, px["XAU"])
plt.title("XAUUSD Price (Weekly Close)")
plt.xlabel("Date"); plt.ylabel("Price")
plt.show()

plt.figure()
plt.plot(px.index, px["XAG"])
plt.title("XAGUSD Price (Weekly Close)")
plt.xlabel("Date"); plt.ylabel("Price")
plt.show()

plt.figure()
plt.plot(beta_used.index, beta_used)
plt.title("Beta used (r_xag ~ beta*r_xau)")
plt.xlabel("Date"); plt.ylabel("beta")
plt.show()

plt.figure()
plt.plot(spread.index, spread)
plt.title("Spread (Silver relative to Gold): log(XAG) - beta*log(XAU)")
plt.xlabel("Date"); plt.ylabel("Spread")
plt.show()

plt.figure()
plt.plot(z.index, z)
plt.axhline(Z_L1, linestyle="--")
plt.axhline(Z_L2, linestyle="--")
plt.axhline(Z_L3, linestyle="--")
plt.axhline(Z_EXIT, linestyle="--")
plt.title("Sigma (z) with ladder thresholds (Silver rich => z>0)")
plt.xlabel("Date"); plt.ylabel("z")
plt.show()

plt.figure()
plt.plot(eq_gross.index, eq_gross, label="Gross equity")
plt.plot(eq_net.index, eq_net, label="Net equity")
plt.title("Equity Curve ($12k start)  Gross vs Net")
plt.xlabel("Date"); plt.ylabel("USD")
plt.legend()
plt.show()

plt.figure()
plt.plot(roll_acf1.index, roll_acf1)
plt.title("Rolling ACF(1) on spread (156w)")
plt.xlabel("Date"); plt.ylabel("ACF1")
plt.show()

plt.figure()
plt.plot(roll_vol.index, roll_vol)
plt.title("Rolling spread volatility (std, 156w)")
plt.xlabel("Date"); plt.ylabel("Std(spread)")
plt.show()

plt.figure()
plt.plot(roll_ent.index, roll_ent)
plt.title("Rolling Shannon entropy of spread (156w)")
plt.xlabel("Date"); plt.ylabel("Entropy")
plt.show()

plt.figure()
plt.plot(roll_adf_p.index, roll_adf_p, label="ADF p")
plt.axhline(0.10, linestyle="--")
plt.title("Rolling ADF p-value on spread (156w)")
plt.xlabel("Date"); plt.ylabel("p-value")
plt.show()




```
