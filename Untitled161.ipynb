```python
# ============================================================
# NFCI × VIX FULL INSTITUTIONAL PROJECT (NO SHORTCUTS)
# DAILY + WEEKLY (AUTO-DETECT), REGIMES, SHOCKS, TRANSMISSION,
# PROPER HEDGE-RATIO SPREAD, OU / HALF-LIFE, REVERSION DENSITY,
# WALK-FORWARD REGIME VISUALS, CLEAN MONOCHROME CHARTS
# ============================================================

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import lstsq
from scipy.stats import gaussian_kde
from statsmodels.tsa.stattools import adfuller, grangercausalitytests
from statsmodels.tsa.vector_ar.vecm import coint_johansen

# ----------------------------
# PROFESSIONAL MONOCHROME STYLE
# ----------------------------
plt.style.use("default")
plt.rcParams.update({
    "font.family": "DejaVu Serif",
    "axes.edgecolor": "black",
    "axes.linewidth": 1.2,
    "figure.facecolor": "white",
    "axes.facecolor": "white",
    "xtick.color": "black",
    "ytick.color": "black",
    "text.color": "black",
})

def clean_axes(ax, title=None, xlabel=None, ylabel=None):
    ax.grid(False)
    for spine in ax.spines.values():
        spine.set_visible(True)
    if title: ax.set_title(title, fontsize=12)
    if xlabel: ax.set_xlabel(xlabel)
    if ylabel: ax.set_ylabel(ylabel)

# ----------------------------
# LOAD FILES (YOU UPLOADED)
# ----------------------------
nfci = pd.read_csv("NFCI (3).csv")
vix  = pd.read_csv("VIXCLS (9).csv")

nfci.columns = ["Date", "NFCI"]
vix.columns  = ["Date", "VIX"]

nfci["Date"] = pd.to_datetime(nfci["Date"])
vix["Date"]  = pd.to_datetime(vix["Date"])

nfci = nfci.sort_values("Date").set_index("Date")
vix  = vix.sort_values("Date").set_index("Date")

nfci["NFCI"] = pd.to_numeric(nfci["NFCI"], errors="coerce")
vix["VIX"]   = pd.to_numeric(vix["VIX"], errors="coerce")

base = pd.concat([nfci, vix], axis=1).dropna()

print("Raw Start:", base.index.min())
print("Raw End:", base.index.max())
print("Raw Observations:", len(base))

# ----------------------------
# AUTO-DETECT FREQUENCY + BUILD DAILY/WEEKLY PANELS
# ----------------------------
def median_step_days(idx):
    d = idx.to_series().diff().dropna().dt.days
    return float(d.median()) if len(d) else np.nan

step = median_step_days(base.index)

# If step ~1–2 => daily-ish, else likely weekly already
if step <= 2.5:
    df_daily  = base.copy()
    df_weekly = base.resample("W-FRI").last().dropna()
    freq_note = "Detected DAILY input -> weekly resampled (W-FRI)."
else:
    df_weekly = base.copy()
    # Create a daily panel via forward-fill only if you explicitly need it later
    # (we keep the original for rigor; no artificial interpolation for tests)
    df_daily  = None
    freq_note = f"Detected NON-DAILY input (median step ~{step:.1f} days) -> treating as WEEKLY already."

print("\nFrequency:", freq_note)
print("Weekly Start:", df_weekly.index.min())
print("Weekly End:", df_weekly.index.max())
print("Weekly Observations:", len(df_weekly))

# ----------------------------
# CORE SETTINGS
# ----------------------------
ROLL_W = 52            # rolling window in weeks for percentiles/z
REG_LO, REG_HI = 0.2, 0.8
SHOCK_Q = 0.95
MAX_LAG = 8

df = df_weekly.copy()

# ============================================================
# 1) COINTEGRATION (JOHANSEN) ON LEVELS
# ============================================================
joh = coint_johansen(df[["NFCI", "VIX"]], det_order=0, k_ar_diff=2)
print("\nJohansen Trace Stats:", joh.lr1)
print("Johansen 95% Critical:", joh.cvt[:, 1])

# ============================================================
# 2) PROPER HEDGE RATIO + SPREAD (RESIDUAL)
#    NFCI = alpha + beta*VIX + e  -> Spread = e
# ============================================================
X = np.column_stack([np.ones(len(df)), df["VIX"].values])
y = df["NFCI"].values

alpha_beta = lstsq(X, y, rcond=None)[0]
alpha, beta = float(alpha_beta[0]), float(alpha_beta[1])

df["Spread"] = df["NFCI"] - (alpha + beta * df["VIX"])

print("\nHedge Model: NFCI = alpha + beta*VIX + e")
print("alpha:", alpha)
print("beta :", beta)

# ============================================================
# 3) STATIONARITY (ADF) ON LEVELS + SPREAD
# ============================================================
def adf_p(series):
    return float(adfuller(series.dropna())[1])

print("\nADF p-values")
print("NFCI  :", adf_p(df["NFCI"]))
print("VIX   :", adf_p(df["VIX"]))
print("Spread:", adf_p(df["Spread"]))

# ============================================================
# 4) OU HALF-LIFE (ON SPREAD) + SPREAD Z-SCORE (ROLLING)
# ============================================================
def half_life(series):
    x = series.dropna()
    dx = x.diff().dropna()
    x_lag = x.shift(1).dropna()
    dx = dx.loc[x_lag.index]
    b = lstsq(x_lag.values.reshape(-1,1), dx.values, rcond=None)[0][0]
    return float(-np.log(2)/b) if b < 0 else np.nan

hl_spread = half_life(df["Spread"])
print("\nSpread Half-Life (weeks):", hl_spread)

df["Spread_mu"] = df["Spread"].rolling(ROLL_W).mean()
df["Spread_sd"] = df["Spread"].rolling(ROLL_W).std()
df["Spread_z"]  = (df["Spread"] - df["Spread_mu"]) / df["Spread_sd"]

df = df.dropna()

# ============================================================
# 5) SHOCKS + TRANSMISSION (Δ series, lag corr, Granger)
# ============================================================
df["dNFCI"] = df["NFCI"].diff()
df["dVIX"]  = df["VIX"].diff()

nfci_thr = df["dNFCI"].quantile(SHOCK_Q)
vix_thr  = df["dVIX"].quantile(SHOCK_Q)

df["NFCI_shock"] = df["dNFCI"] > nfci_thr
df["VIX_shock"]  = df["dVIX"]  > vix_thr

print("\nShock Counts (weekly changes > {:.0f}th pct)".format(SHOCK_Q*100))
print("NFCI shocks:", int(df["NFCI_shock"].sum()))
print("VIX shocks :", int(df["VIX_shock"].sum()))

# Lag correlation
lags = range(-MAX_LAG, MAX_LAG+1)
lag_corr = [df["dVIX"].corr(df["dNFCI"].shift(l)) for l in lags]

fig, ax = plt.subplots(figsize=(8.6, 3.8))
ax.plot(list(lags), lag_corr, color="black", linewidth=1.5)
ax.axhline(0, color="black", linewidth=1)
clean_axes(ax, "Lag Correlation: ΔVIX vs ΔNFCI", "Lag (weeks)", "Correlation")
plt.tight_layout()
plt.show()

# Granger (quiet; you can inspect p-values if needed)
def granger_min_p(y_col, x_col, maxlag=8):
    # Tests: does x help predict y?
    data = df[[y_col, x_col]].dropna()
    res = grangercausalitytests(data, maxlag=maxlag, verbose=False)
    pvals = {lag: res[lag][0]["ssr_ftest"][1] for lag in res}
    return pvals, min(pvals.values())

p_vix_to_nfci, min_vix_to_nfci = granger_min_p("dNFCI", "dVIX", MAX_LAG)
p_nfci_to_vix, min_nfci_to_vix = granger_min_p("dVIX", "dNFCI", MAX_LAG)

print("\nGranger (min p across lags 1..{}):".format(MAX_LAG))
print("VIX → NFCI (Δ):", min_vix_to_nfci)
print("NFCI → VIX (Δ):", min_nfci_to_vix)

# ============================================================
# 6) STRUCTURAL REGIMES (NFCI & VIX) + SPREAD REGIME
#    based on rolling percentiles (20/80)
# ============================================================
def rolling_pct_last(s, w):
    return s.rolling(w).apply(lambda x: pd.Series(x).rank(pct=True).iloc[-1])

df["NFCI_pct"]  = rolling_pct_last(df["NFCI"],  ROLL_W)
df["VIX_pct"]   = rolling_pct_last(df["VIX"],   ROLL_W)
df["Spr_pct"]   = rolling_pct_last(df["Spread"],ROLL_W)

def bucket(p, lo=REG_LO, hi=REG_HI):
    if p < lo: return "Low"
    if p > hi: return "High"
    return "Neutral"

df["NFCI_reg"] = df["NFCI_pct"].apply(bucket)
df["VIX_reg"]  = df["VIX_pct"].apply(bucket)

# Spread regime is "Undervalued/Overvalued" around its own distribution
def spr_bucket(p, lo=REG_LO, hi=REG_HI):
    if p < lo: return "Undervalued"
    if p > hi: return "Overvalued"
    return "Neutral"

df["Spread_reg"] = df["Spr_pct"].apply(spr_bucket)

print("\nRegime counts (NFCI × VIX):")
print(pd.crosstab(df["NFCI_reg"], df["VIX_reg"]))

print("\nSpread regime counts:")
print(df["Spread_reg"].value_counts())

# ============================================================
# 7) WALK-FORWARD: NFCI_reg(t) -> Prob[VIX_reg(t+h)=High]
#    plus visual "walk-forward markers" on chart
# ============================================================
def walkforward_prob(source_reg, target_reg, target_state="High", horizons=(1,2,4,8)):
    out = {}
    for h in horizons:
        future = target_reg.shift(-h)
        tab = pd.crosstab(source_reg, future, normalize="index")
        out[h] = tab[target_state] if target_state in tab.columns else pd.Series(np.nan, index=tab.index)
    return pd.DataFrame(out)

wf = walkforward_prob(df["NFCI_reg"], df["VIX_reg"], "High", horizons=(1,2,4,8))
print("\nWalk-forward P(VIX High at t+h | NFCI regime at t):")
print(wf)

# ============================================================
# 8) REVERSION DENSITY (THE THING YOU ASKED FOR)
#    - For extreme spread z, measure distribution of future z-change
#    - And probability of reverting inside |z|<1 within h weeks
# ============================================================
def forward_delta(series, h):
    return series.shift(-h) - series

ext_hi = df["Spread_z"] >= 2.0
ext_lo = df["Spread_z"] <= -2.0

horizons = [1,2,4,8,13,26]  # weeks (up to ~6 months)
revert_probs = []

for h in horizons:
    z_fwd = df["Spread_z"].shift(-h)
    # "reverted" = inside ±1 at horizon h
    reverted = (z_fwd.abs() <= 1.0)

    p_hi = float(reverted[ext_hi].mean()) if ext_hi.sum() else np.nan
    p_lo = float(reverted[ext_lo].mean()) if ext_lo.sum() else np.nan
    revert_probs.append([h, p_hi, p_lo])

revert_probs = pd.DataFrame(revert_probs, columns=["h_weeks", "P_revert_from_z>=2", "P_revert_from_z<=-2"])
print("\nReversion probability into |z|<=1 after h weeks (conditional on starting extreme):")
print(revert_probs)

# Density of forward z-change conditional on extremes
fig, ax = plt.subplots(figsize=(8.6, 3.8))
xs = np.linspace(-4, 4, 400)

for label, mask, col in [
    ("Start z>=2 : Δz(4w)", ext_hi, "black"),
    ("Start z<=-2: Δz(4w)", ext_lo, "grey"),
]:
    dz = forward_delta(df["Spread_z"], 4)[mask].dropna()
    if len(dz) > 30:
        kde = gaussian_kde(dz.values)
        ax.plot(xs, kde(xs), color=col, linewidth=1.6, label=label)

ax.axvline(0, color="black", linewidth=1)
clean_axes(ax, "Reversion Density: Δ(Spread z) over 4 weeks | extreme start", "Δz over 4 weeks", "Density")
ax.legend(frameon=False)
plt.tight_layout()
plt.show()

# Reversion probability curve
fig, ax = plt.subplots(figsize=(8.6, 3.8))
ax.plot(revert_probs["h_weeks"], revert_probs["P_revert_from_z>=2"], color="black", linewidth=1.8, label="Start z>=2")
ax.plot(revert_probs["h_weeks"], revert_probs["P_revert_from_z<=-2"], color="grey", linewidth=1.8, label="Start z<=-2")
ax.set_ylim(0, 1)
clean_axes(ax, "Mean Reversion: P(|z|<=1) after h weeks (conditional on extremes)", "h (weeks)", "Probability")
ax.legend(frameon=False)
plt.tight_layout()
plt.show()

# ============================================================
# 9) INSTITUTIONAL LINE CHARTS YOU ASKED FOR:
#    - NFCI with rolling 20/80 regime bands + shaded High/Low
#    - VIX with rolling 20/80 regime bands + shaded High/Low
#    - Spread_z with ±2 boundaries + shaded extreme regimes
#    - Walk-forward markers: mark times when NFCI is High and show if VIX is High in +4w (visual)
# ============================================================

def plot_series_with_bands(idx, series, low_band, high_band, title, ylab):
    fig, ax = plt.subplots(figsize=(12.5, 4.0))
    ax.plot(idx, series, color="black", linewidth=1.2)
    ax.plot(idx, high_band, color="grey", linestyle="--", linewidth=1.0)
    ax.plot(idx, low_band,  color="grey", linestyle="--", linewidth=1.0)

    # Shade High region (above high_band) and Low region (below low_band)
    ax.fill_between(idx, series, high_band, where=(series >= high_band), color="grey", alpha=0.25)
    ax.fill_between(idx, series, low_band,  where=(series <= low_band),  color="grey", alpha=0.12)

    clean_axes(ax, title, None, ylab)
    plt.tight_layout()
    plt.show()

# Rolling 20/80 bands (weekly)
df["NFCI_q20"] = df["NFCI"].rolling(ROLL_W).quantile(REG_LO)
df["NFCI_q80"] = df["NFCI"].rolling(ROLL_W).quantile(REG_HI)
df["VIX_q20"]  = df["VIX"].rolling(ROLL_W).quantile(REG_LO)
df["VIX_q80"]  = df["VIX"].rolling(ROLL_W).quantile(REG_HI)

plot_series_with_bands(df.index, df["NFCI"], df["NFCI_q20"], df["NFCI_q80"],
                       "NFCI (Level) with Rolling 20/80 Regime Bands", "NFCI")

plot_series_with_bands(df.index, df["VIX"], df["VIX_q20"], df["VIX_q80"],
                       "VIX (Level) with Rolling 20/80 Regime Bands", "VIX")

# Spread z chart with ±2 and shaded extremes
fig, ax = plt.subplots(figsize=(12.5, 4.0))
ax.plot(df.index, df["Spread_z"], color="black", linewidth=1.2)
ax.axhline(0, color="black", linewidth=1.0)
ax.axhline(2, color="grey", linestyle="--", linewidth=1.0)
ax.axhline(-2, color="grey", linestyle="--", linewidth=1.0)

ax.fill_between(df.index, df["Spread_z"], 2, where=(df["Spread_z"] >= 2), color="grey", alpha=0.25)
ax.fill_between(df.index, df["Spread_z"], -2, where=(df["Spread_z"] <= -2), color="grey", alpha=0.12)

clean_axes(ax, "Spread Z-Score (Residual of NFCI ~ VIX) with ±2 Extremes", None, "Spread_z")
plt.tight_layout()
plt.show()

# Walk-forward visual marker:
# Mark dates where NFCI is High, and indicate whether VIX is High in +4 weeks.
marker_dates = df.index[df["NFCI_reg"] == "High"]
vix_high_t4  = (df["VIX_reg"].shift(-4) == "High")

fig, ax = plt.subplots(figsize=(12.5, 4.0))
ax.plot(df.index, df["VIX"], color="black", linewidth=1.1)
ax.plot(df.index, df["VIX_q80"], color="grey", linestyle="--", linewidth=1.0)
# Grey vertical ticks when NFCI high; darker tick when VIX will be high at +4w
ax.vlines(marker_dates, ymin=df["VIX"].min(), ymax=df["VIX"].min() + (df["VIX"].max()-df["VIX"].min())*0.08,
          color="grey", linewidth=0.8, alpha=0.5)
ax.vlines(df.index[(df["NFCI_reg"]=="High") & vix_high_t4], ymin=df["VIX"].min(), ymax=df["VIX"].min() + (df["VIX"].max()-df["VIX"].min())*0.12,
          color="black", linewidth=1.1, alpha=0.9)

clean_axes(ax, "Walk-Forward Visual: NFCI High at t (ticks) | Black ticks = VIX High at t+4w", None, "VIX")
plt.tight_layout()
plt.show()

# ============================================================
# 10) SPREAD DENSITY + SPREAD_Z DENSITY (MONOCHROME)
# ============================================================
def plot_kde(series, title, xlabel):
    x = series.dropna().values
    xs = np.linspace(np.nanpercentile(x, 0.5), np.nanpercentile(x, 99.5), 400)
    kde = gaussian_kde(x)
    fig, ax = plt.subplots(figsize=(8.6, 3.8))
    ax.plot(xs, kde(xs), color="black", linewidth=1.6)
    clean_axes(ax, title, xlabel, "Density")
    plt.tight_layout()
    plt.show()

plot_kde(df["Spread"],   "Spread Density (Residual of NFCI ~ VIX)", "Spread")
plot_kde(df["Spread_z"], "Spread Z-Score Density", "Spread_z")

# ============================================================
# 11) REGIME DURATION STATS (YOU ASKED FOR INDICATORS + TESTS)
# ============================================================
df["Spr_change"] = df["Spread_reg"] != df["Spread_reg"].shift(1)
df["Spr_block"]  = df["Spr_change"].cumsum()
dur = df.groupby(["Spr_block","Spread_reg"]).size()
avg_dur = dur.groupby(level=1).mean()
med_dur = dur.groupby(level=1).median()

print("\nSpread regime duration (weeks):")
print(pd.DataFrame({"avg_weeks": avg_dur, "median_weeks": med_dur}))

# ============================================================
# 12) DAILY PANEL (ONLY IF YOUR INPUT REALLY IS DAILY)
#     We run the same core spread + tests on DAILY too, without
#     inventing data. If df_daily exists, do it.
# ============================================================
if df_daily is not None:
    ddf = df_daily.copy()

    # Hedge model on daily
    Xd = np.column_stack([np.ones(len(ddf)), ddf["VIX"].values])
    yd = ddf["NFCI"].values
    a_b = lstsq(Xd, yd, rcond=None)[0]
    a_d, b_d = float(a_b[0]), float(a_b[1])
    ddf["Spread"] = ddf["NFCI"] - (a_d + b_d * ddf["VIX"])

    # Daily ADF
    print("\n[DAILY] Hedge Model: NFCI = alpha + beta*VIX + e")
    print("[DAILY] alpha:", a_d)
    print("[DAILY] beta :", b_d)
    print("\n[DAILY] ADF p-values")
    print("[DAILY] NFCI  :", adf_p(ddf["NFCI"]))
    print("[DAILY] VIX   :", adf_p(ddf["VIX"]))
    print("[DAILY] Spread:", adf_p(ddf["Spread"]))

    # Daily OU half-life in DAYS (same estimator, different unit)
    hl_days = half_life(ddf["Spread"])
    print("\n[DAILY] Spread Half-Life (days):", hl_days)

    # Daily spread z-score (rolling 252 trading days)
    dW = 252
    ddf["Spread_z"] = (ddf["Spread"] - ddf["Spread"].rolling(dW).mean()) / ddf["Spread"].rolling(dW).std()
    ddf = ddf.dropna()

    # Daily spread_z with ±2 (monochrome)
    fig, ax = plt.subplots(figsize=(12.5, 4.0))
    ax.plot(ddf.index, ddf["Spread_z"], color="black", linewidth=1.0)
    ax.axhline(0, color="black", linewidth=1.0)
    ax.axhline(2, color="grey", linestyle="--", linewidth=1.0)
    ax.axhline(-2, color="grey", linestyle="--", linewidth=1.0)
    ax.fill_between(ddf.index, ddf["Spread_z"], 2, where=(ddf["Spread_z"] >= 2), color="grey", alpha=0.22)
    ax.fill_between(ddf.index, ddf["Spread_z"], -2, where=(ddf["Spread_z"] <= -2), color="grey", alpha=0.10)
    clean_axes(ax, "[DAILY] Spread Z-Score with ±2 Extremes (252d rolling)", None, "Spread_z")
    plt.tight_layout()
    plt.show()
else:
    print("\nDAILY module skipped: your input is not daily frequency (already weekly).")


```
