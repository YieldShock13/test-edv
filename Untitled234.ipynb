```python
# ============================================================
# RSPS - RSPD Spread Analysis (CLEANED)
# Full sample: 2015 - Present
# Removes corrupted rows / fake spikes
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from statsmodels.tsa.stattools import adfuller
import warnings
warnings.filterwarnings("ignore")

# ============================================================
# 1. LOAD + CLEAN DATA
# ============================================================

rsps = pd.read_csv("RSPS Historical Data (1).csv")
rspd = pd.read_csv("RSPD ETF Stock Price History.csv")

# ----------------------------
# Clean RSPS
# ----------------------------

rsps = rsps.dropna(subset=["Vol."])

rsps["Date"] = pd.to_datetime(rsps["Date"])

rsps["Price"] = (
    rsps["Price"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .astype(float)
)

# Remove absurd corrupted values
rsps = rsps[
    (rsps["Price"] > 1) &
    (rsps["Price"] < 1000)
]

# Remove extreme daily jumps
rsps["ret"] = rsps["Price"].pct_change()

rsps = rsps[
    rsps["ret"].abs() < 0.25
]

# ----------------------------
# Clean RSPD
# ----------------------------

rspd = rspd.dropna(subset=["Vol."])

rspd["Date"] = pd.to_datetime(rspd["Date"])

rspd["Price"] = (
    rspd["Price"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .astype(float)
)

rspd = rspd[
    (rspd["Price"] > 1) &
    (rspd["Price"] < 1000)
]

rspd["ret"] = rspd["Price"].pct_change()

rspd = rspd[
    rspd["ret"].abs() < 0.25
]

# ============================================================
# 2. PREP SERIES
# ============================================================

rsps = rsps[["Date", "Price"]].rename(columns={"Price": "RSPS"})
rspd = rspd[["Date", "Price"]].rename(columns={"Price": "RSPD"})

df = pd.merge(rsps, rspd, on="Date", how="inner")

df = df.sort_values("Date").reset_index(drop=True)

# Full period
df = df[df["Date"] >= "2015-01-01"].copy()

# Normal spread
df["Spread"] = df["RSPS"] - df["RSPD"]

# ============================================================
# 3. CHART STYLE
# ============================================================

plt.style.use("seaborn-v0_8-darkgrid")

plt.rcParams["font.family"] = "DejaVu Sans"
plt.rcParams["axes.titleweight"] = "bold"
plt.rcParams["axes.titlesize"] = 15
plt.rcParams["axes.labelsize"] = 11
plt.rcParams["xtick.labelsize"] = 10
plt.rcParams["ytick.labelsize"] = 10

# ============================================================
# 4. PRICE SERIES
# ============================================================

fig, ax = plt.subplots(figsize=(13, 6))

ax.plot(df["Date"], df["RSPS"], linewidth=2.0, label="RSPS")
ax.plot(df["Date"], df["RSPD"], linewidth=2.0, label="RSPD")

ax.set_title("RSPS vs RSPD Price History")
ax.set_ylabel("Price")
ax.set_xlabel("")

ax.legend(frameon=True)

ax.text(
    0.01,
    0.02,
    "Pearl Gray Equity and Research",
    transform=ax.transAxes,
    fontsize=10,
    alpha=0.55
)

plt.tight_layout()
plt.show()

# ============================================================
# 5. NORMAL SPREAD
# ============================================================

spread_mean = df["Spread"].mean()
latest_spread = df["Spread"].iloc[-1]

fig, ax = plt.subplots(figsize=(13, 6))

ax.plot(
    df["Date"],
    df["Spread"],
    linewidth=2.2,
    label="RSPS - RSPD Spread"
)

ax.axhline(
    spread_mean,
    linestyle="--",
    linewidth=1.4,
    label=f"Mean: {spread_mean:.2f}"
)

ax.scatter(
    df["Date"].iloc[-1],
    latest_spread,
    s=60,
    zorder=5
)

ax.annotate(
    f"{latest_spread:.2f}",
    xy=(df["Date"].iloc[-1], latest_spread),
    xytext=(10, 10),
    textcoords="offset points",
    fontsize=11,
    fontweight="bold"
)

ax.set_title("RSPS - RSPD Normal Price Spread")
ax.set_ylabel("Price Spread")
ax.set_xlabel("")

ax.legend(frameon=True)

ax.text(
    0.01,
    0.02,
    "Pearl Gray Equity and Research",
    transform=ax.transAxes,
    fontsize=10,
    alpha=0.55
)

plt.tight_layout()
plt.show()

# ============================================================
# 6. ADF TEST
# ============================================================

x = df["Spread"].values

adf_stat, adf_p, *_ = adfuller(x)

print("ADF on spread")
print("ADF statistic:", round(adf_stat, 4))
print("ADF p-value:", round(adf_p, 4))

# ============================================================
# 7. OU MODEL
# ============================================================

dt = 1 / 252

x_t = x[:-1]
x_next = x[1:]

def ou_neg_log_likelihood(params):

    kappa, mu, sigma = params

    if kappa <= 0 or sigma <= 0:
        return 1e10

    exp_term = np.exp(-kappa * dt)

    mean = mu + (x_t - mu) * exp_term

    variance = (
        (sigma**2 / (2 * kappa))
        * (1 - np.exp(-2 * kappa * dt))
    )

    ll = -0.5 * np.sum(
        np.log(2 * np.pi * variance)
        + ((x_next - mean) ** 2) / variance
    )

    return -ll

initial_guess = [
    1.0,
    np.mean(x),
    np.std(np.diff(x)) * np.sqrt(252)
]

result = minimize(
    ou_neg_log_likelihood,
    initial_guess,
    bounds=[
        (1e-6, None),
        (None, None),
        (1e-6, None)
    ]
)

kappa, mu, sigma = result.x

half_life_days = (
    np.log(2) / kappa
) * 252

ou_std = sigma / np.sqrt(2 * kappa)

df["OU_Z"] = (
    df["Spread"] - mu
) / ou_std

print("\nOU RESULTS")
print("Kappa:", round(kappa, 4))
print("Mean:", round(mu, 4))
print("Sigma:", round(sigma, 4))
print("Half-life:", round(half_life_days, 2), "days")
print("Latest z-score:", round(df["OU_Z"].iloc[-1], 2))

# ============================================================
# 8. OU Z-SCORE CHART
# ============================================================

latest_z = df["OU_Z"].iloc[-1]

fig, ax = plt.subplots(figsize=(13, 6))

ax.plot(
    df["Date"],
    df["OU_Z"],
    linewidth=2.0,
    label="OU Z-score"
)

ax.axhline(0, linestyle="--", linewidth=1.2)
ax.axhline(1, linestyle=":", linewidth=1.2)
ax.axhline(-1, linestyle=":", linewidth=1.2)
ax.axhline(2, linestyle=":", linewidth=1.2)
ax.axhline(-2, linestyle=":", linewidth=1.2)

ax.scatter(
    df["Date"].iloc[-1],
    latest_z,
    s=60,
    zorder=5
)

ax.annotate(
    f"{latest_z:.2f}σ",
    xy=(df["Date"].iloc[-1], latest_z),
    xytext=(10, 10),
    textcoords="offset points",
    fontsize=11,
    fontweight="bold"
)

ax.set_title("RSPS - RSPD Spread: OU Z-score")
ax.set_ylabel("Z-score")
ax.set_xlabel("")

ax.legend(frameon=True)

ax.text(
    0.01,
    0.02,
    "Pearl Gray Equity and Research",
    transform=ax.transAxes,
    fontsize=10,
    alpha=0.55
)

plt.tight_layout()
plt.show()

# ============================================================
# 9. ROLLING AUTOCORRELATION
# ============================================================

window = 60

df["Rolling_ACF_1"] = (
    df["Spread"]
    .rolling(window)
    .apply(
        lambda s: pd.Series(s).autocorr(lag=1),
        raw=False
    )
)

latest_acf = (
    df["Rolling_ACF_1"]
    .dropna()
    .iloc[-1]
)

fig, ax = plt.subplots(figsize=(13, 6))

ax.plot(
    df["Date"],
    df["Rolling_ACF_1"],
    linewidth=2.0,
    label="60d rolling lag-1 autocorrelation"
)

ax.axhline(
    0,
    linestyle="--",
    linewidth=1.2
)

ax.scatter(
    df["Date"].iloc[-1],
    latest_acf,
    s=60,
    zorder=5
)

ax.annotate(
    f"{latest_acf:.2f}",
    xy=(df["Date"].iloc[-1], latest_acf),
    xytext=(10, 10),
    textcoords="offset points",
    fontsize=11,
    fontweight="bold"
)

ax.set_title("RSPS - RSPD Spread: Rolling Autocorrelation")
ax.set_ylabel("Autocorrelation")
ax.set_xlabel("")

ax.legend(frameon=True)

ax.text(
    0.01,
    0.02,
    "Pearl Gray Equity and Research",
    transform=ax.transAxes,
    fontsize=10,
    alpha=0.55
)

plt.tight_layout()
plt.show()

# ============================================================
# 10. ROLLING ENTROPY
# ============================================================

def shannon_entropy(values, bins=20):

    counts, _ = np.histogram(values, bins=bins)

    probs = counts / counts.sum()

    probs = probs[probs > 0]

    return -np.sum(probs * np.log(probs))

df["Rolling_Entropy"] = (
    df["Spread"]
    .rolling(window)
    .apply(
        lambda s: shannon_entropy(s, bins=20),
        raw=False
    )
)

latest_entropy = (
    df["Rolling_Entropy"]
    .dropna()
    .iloc[-1]
)

fig, ax = plt.subplots(figsize=(13, 6))

ax.plot(
    df["Date"],
    df["Rolling_Entropy"],
    linewidth=2.0,
    label="60d Shannon entropy"
)

ax.scatter(
    df["Date"].iloc[-1],
    latest_entropy,
    s=60,
    zorder=5
)

ax.annotate(
    f"{latest_entropy:.2f}",
    xy=(df["Date"].iloc[-1], latest_entropy),
    xytext=(10, 10),
    textcoords="offset points",
    fontsize=11,
    fontweight="bold"
)

ax.set_title("RSPS - RSPD Spread: Rolling Entropy")
ax.set_ylabel("Entropy")
ax.set_xlabel("")

ax.legend(frameon=True)

ax.text(
    0.01,
    0.02,
    "Pearl Gray Equity and Research",
    transform=ax.transAxes,
    fontsize=10,
    alpha=0.55
)

plt.tight_layout()
plt.show()
```
