```python
# === HY OAS vs IG OAS – Quarterly (label at last actual date) + Z-Score ===
# Units auto-detected (fraction/percent/bps) and normalized to bps.
# Colab-ready: just set the two filenames you already have in /content.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ---------------- CONFIG ----------------
HY_CSV_PATH = "BAMLH0A0HYM2 (7).csv"   # <- update if different
IG_CSV_PATH = "BAMLC0A0CM.csv"         # <- update if different

SERIES_NAME_HY = "HY OAS"
SERIES_NAME_IG = "IG OAS"

SPREAD_NAME    = "HY – IG OAS (bps)"
Z_NAME         = "Z-Score (HY–IG)"
WATERMARK_TEXT = "Orca’s Spread"

# Quarterly aggregation: False = quarterly average; True = last value in quarter
QUARTER_USE_LAST = False

# Z-score: None = full-sample; or int for rolling window (quarters)
ROLLING_WINDOW = None

# Appearance
DPI = 160
FONT_FAMILY = "DejaVu Sans"
plt.rcParams.update({
    "font.family": FONT_FAMILY,
    "figure.facecolor": "black",
    "axes.facecolor": "black",
    "savefig.facecolor": "black",
    "text.color": "white",
    "axes.labelcolor": "white",
    "xtick.color": "white",
    "ytick.color": "white",
    "axes.grid": True,
    "grid.color": "#555555",
    "grid.alpha": 0.25
})
# ---------------------------------------


def read_oas_csv(path):
    df = pd.read_csv(path)
    # auto-detect date col
    date_col = None
    for c in df.columns:
        try:
            pd.to_datetime(df[c])
            date_col = c
            break
        except Exception:
            continue
    if date_col is None:
        raise ValueError(f"No parseable date column found in {path}.")

    # pick last numeric col (common for these ICE/BofA extracts)
    numeric_cols = [c for c in df.columns if pd.api.types.is_numeric_dtype(df[c]) and c != date_col]
    if not numeric_cols:
        for c in df.columns:
            if c != date_col:
                df[c] = pd.to_numeric(df[c].astype(str).str.replace(",",""), errors="coerce")
        numeric_cols = [c for c in df.columns if pd.api.types.is_numeric_dtype(df[c]) and c != date_col]
        if not numeric_cols:
            raise ValueError(f"No numeric value column found in {path}.")
    value_col = numeric_cols[-1]

    s = pd.Series(
        pd.to_numeric(df[value_col], errors="coerce").values,
        index=pd.to_datetime(df[date_col]),
        name=value_col
    ).dropna().sort_index()

    # --- robust unit detection ---
    p95 = float(s.quantile(0.95))
    if p95 <= 1.0:
        # fraction (e.g., 0.035) -> bps
        s = s * 10000.0
    elif p95 <= 20.0:
        # percent (e.g., 3.5) -> bps
        s = s * 100.0
    else:
        # already in bps (e.g., 350)
        pass

    return s


def quarterly_label_last_actual(series: pd.Series, use_last: bool) -> pd.Series:
    """
    Aggregate by quarter but timestamp each quarter at the LAST ACTUAL DATE
    present in that quarter (nearest available), not the calendar quarter-end.
    """
    q = series.copy()
    periods = q.index.to_period("Q")
    last_dates = pd.Series(q.index).groupby(periods).max()

    if use_last:
        idx_last = q.groupby(periods).apply(lambda s: s.index.max())
        vals = q.loc[idx_last.values].values
        out = pd.Series(vals, index=last_dates.values, name=series.name)
    else:
        means = q.groupby(periods).mean()
        out = pd.Series(means.values, index=last_dates.values, name=series.name)

    return out.sort_index().dropna()


def zscore(series: pd.Series, rolling_window=None) -> pd.Series:
    if rolling_window is None:
        mu = series.mean()
        sd = series.std(ddof=1)
        return (series - mu) / sd
    mu = series.rolling(rolling_window, min_periods=max(4, rolling_window//2)).mean()
    sd = series.rolling(rolling_window, min_periods=max(4, rolling_window//2)).std(ddof=1)
    return (series - mu) / sd


def add_watermark(fig, text=WATERMARK_TEXT, alpha=0.25):
    fig.text(0.995, 0.02, text, fontsize=10, color="white",
             ha="right", va="bottom", alpha=alpha)


# --- Load & align daily ---
hy_daily = read_oas_csv(HY_CSV_PATH).rename(SERIES_NAME_HY)
ig_daily = read_oas_csv(IG_CSV_PATH).rename(SERIES_NAME_IG)
daily = pd.concat([hy_daily, ig_daily], axis=1).dropna().sort_index()

# --- Quarterly with labels at last actual date ---
hy_q = quarterly_label_last_actual(daily[SERIES_NAME_HY], use_last=QUARTER_USE_LAST)
ig_q = quarterly_label_last_actual(daily[SERIES_NAME_IG], use_last=QUARTER_USE_LAST)
df_q = pd.concat([hy_q, ig_q], axis=1).dropna().sort_index()

# --- Spread & Z ---
spread_q = (df_q[SERIES_NAME_HY] - df_q[SERIES_NAME_IG]).rename(SPREAD_NAME)
z = zscore(spread_q, rolling_window=ROLLING_WINDOW).rename(Z_NAME)

# --- Plot 1: Spread ---
fig1, ax1 = plt.subplots(figsize=(11,4), dpi=DPI)
ax1.plot(spread_q.index, spread_q.values, linewidth=2)
ax1.set_title("HY – IG OAS Spread (Quarterly, labeled at last actual date)", pad=12)
ax1.set_ylabel("Basis Points")
ax1.set_xlabel("Date")
ax1.axhline(0, linestyle="--", alpha=0.5)

last_dt = spread_q.index[-1]
last_val = spread_q.iloc[-1]
ax1.annotate(f"{last_val:.0f} bps ({last_dt.date()})",
             xy=(last_dt, last_val), xytext=(10,0),
             textcoords="offset points", va="center")
add_watermark(fig1)
plt.tight_layout()
plt.show()

# --- Plot 2: Z-Score ---
fig2, ax2 = plt.subplots(figsize=(11,4), dpi=DPI)
ax2.plot(z.index, z.values, linewidth=2)
ax2.set_title("HY–IG OAS Spread – Z-Score (Quarterly, labeled at last actual date)", pad=12)
ax2.set_ylabel("Z")
ax2.set_xlabel("Date")
ax2.axhline(0, linestyle="--", alpha=0.5)

for thr, ls in [(1,"-."),(2,":")]:
    ax2.axhline(thr, linestyle=ls, alpha=0.6)
    ax2.axhline(-thr, linestyle=ls, alpha=0.6)

ax2.fill_between(z.index, 1, z, where=z>=1, alpha=0.12)
ax2.fill_between(z.index, 2, z, where=z>=2, alpha=0.18)
ax2.fill_between(z.index, -1, z, where=z<=-1, alpha=0.12)
ax2.fill_between(z.index, -2, z, where=z<=-2, alpha=0.18)

ax2.annotate(f"{z.iloc[-1]:.2f} σ ({z.index[-1].date()})",
             xy=(z.index[-1], z.iloc[-1]), xytext=(10,0),
             textcoords="offset points", va="center")
add_watermark(fig2)
plt.tight_layout()
plt.show()

# --- Save outputs ---
spread_q.to_frame().to_csv("quarterly_spread_hy_minus_ig_nearest_date.csv")
z.to_frame().to_csv("quarterly_spread_zscore_nearest_date.csv")



```
