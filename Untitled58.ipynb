```python
# --- UK–US CDS Basis (Monthly) — Robust + Interpolated + Medians & Labels ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import os

# ===================== CONFIG =====================
CANDIDATES = {
    "UK_CDS": [
        "/content/GBGV10YEUAC=R Overview (1).csv",
        "/content/GBGV10YEUAC=R Overview.csv",
    ],
    "US_CDS": [
        "/content/USGV10YUSAB=R Overview (4).csv",
        "/content/USGV10YUSAB=R Overview (3).csv",
        "/content/USGV10YEUAC=R Overview.csv",
    ],
    "UK_10Y": [
        "/content/United Kingdom 10-Year Bond Yield Historical Data (8).csv",
        "/content/United Kingdom 10-Year Bond Yield Historical Data (7).csv",
    ],
    "US_10Y": [
        "/content/United States 10-Year Bond Yield Historical Data (6).csv",
        "/content/United States 10-Year Bond Yield Historical Data (5).csv",
    ],
}

START = "2009-08-01"
END   = "2025-08-01"

OUT_RAW = "/content/cds_basis_uk_us_monthly_raw.csv"
OUT_INT = "/content/cds_basis_uk_us_monthly_interpolated.csv"

# ===================== HELPERS =====================
def pick_first_existing(paths):
    for p in paths:
        if os.path.exists(p):
            return p
    raise FileNotFoundError(f"None of the expected files found: {paths}")

def _clean_numeric(series: pd.Series) -> pd.Series:
    s = (series.astype(str)
               .str.replace(r"[,\u00A0\s]", "", regex=True)
               .str.replace("%", "", regex=True))
    return pd.to_numeric(s, errors="coerce")

def _guess_value_col(df: pd.DataFrame) -> str:
    preferred = ["Close","Last","Price","Value","Mid","CDS","Yield","Rate"]
    for p in preferred:
        if p in df.columns:
            return p
    for c in df.columns:
        try:
            pd.to_numeric(df[c].astype(str).str.replace(r"[,\s%]", "", regex=True), errors="raise")
            return c
        except Exception:
            continue
    raise ValueError("No numeric value column detected.")

def _find_date_col(df: pd.DataFrame) -> str:
    for c in df.columns:
        try:
            pd.to_datetime(df[c], errors="raise", dayfirst=True); return c
        except Exception:
            try:
                pd.to_datetime(df[c], errors="raise", dayfirst=False); return c
            except Exception:
                continue
    raise ValueError("No date-like column detected.")

def _parse_dates_best(df: pd.DataFrame, date_col: str) -> pd.Series:
    d1 = pd.to_datetime(df[date_col], errors="coerce", dayfirst=True)
    d2 = pd.to_datetime(df[date_col], errors="coerce", dayfirst=False)
    def score(d): return (d.notna().sum(), pd.Timestamp.min if d.dropna().empty else d.max())
    return d1 if score(d1) > score(d2) else d2

def load_series_generic(path: str) -> pd.Series:
    df = pd.read_csv(path)
    date_col  = _find_date_col(df)
    value_col = _guess_value_col(df)
    dates = _parse_dates_best(df, date_col)
    vals  = _clean_numeric(df[value_col])
    tmp = pd.DataFrame({"Date": dates, "Val": vals}).dropna(subset=["Date","Val"])
    tmp = tmp.sort_values("Date").set_index("Date")
    s = tmp["Val"].astype(float)
    s.name = Path(path).stem
    return s

def ensure_bps(s: pd.Series) -> pd.Series:
    if s.dropna().empty: return s
    pct95 = np.nanpercentile(s.dropna().values, 95)
    return s * 100.0 if pct95 <= 30 else s

def add_median_line(ax, series, label_suffix="median"):
    m = float(np.nanmedian(series))
    ax.axhline(m, linestyle="--", linewidth=1)
    ax.text(ax.get_xlim()[1], m, f"  {label_suffix}: {m:.2f}",
            va="center", ha="left", fontsize=9, clip_on=True)

def label_last_point(ax, x, y, text):
    ax.annotate(text, xy=(x, y), xytext=(5, 0),
                textcoords="offset points", fontsize=9, va="center")

# ===================== LOAD =====================
UK_CDS_FILE = pick_first_existing(CANDIDATES["UK_CDS"])
US_CDS_FILE = pick_first_existing(CANDIDATES["US_CDS"])
UK_10Y_FILE = pick_first_existing(CANDIDATES["UK_10Y"])
US_10Y_FILE = pick_first_existing(CANDIDATES["US_10Y"])

uk_cds = ensure_bps(load_series_generic(UK_CDS_FILE))
us_cds = ensure_bps(load_series_generic(US_CDS_FILE))
uk_10y = ensure_bps(load_series_generic(UK_10Y_FILE))
us_10y = ensure_bps(load_series_generic(US_10Y_FILE))

print("Daily coverage (min → max):")
for name, ser in [("UK_CDS",uk_cds),("US_CDS",us_cds),("UK_10Y",uk_10y),("US_10Y",us_10y)]:
    print(f"  {name}: {ser.index.min().date()} → {ser.index.max().date()} (n={ser.size})")

# ===================== MONTHLY =====================
uk_cds_m = uk_cds.resample("M").mean()
us_cds_m = us_cds.resample("M").mean()
uk_10y_m = uk_10y.resample("M").mean()
us_10y_m = us_10y.resample("M").mean()

m_index = pd.date_range(start=START, end=END, freq="M")
df = pd.concat(
    [uk_cds_m.rename("UK_CDS_bp"),
     us_cds_m.rename("US_CDS_bp"),
     uk_10y_m.rename("UK_10Y_bp"),
     us_10y_m.rename("US_10Y_bp")],
    axis=1
).reindex(m_index)

# ---------- RAW BASIS ----------
df_raw = df.copy()
df_raw["CDS_diff_bp"]  = df_raw["UK_CDS_bp"] - df_raw["US_CDS_bp"]
df_raw["Y10_diff_bp"]  = df_raw["UK_10Y_bp"] - df_raw["US_10Y_bp"]
df_raw["CDS_basis_bp"] = df_raw["CDS_diff_bp"] - df_raw["Y10_diff_bp"]
df_raw["basis_z"]      = (df_raw["CDS_basis_bp"] - df_raw["CDS_basis_bp"].mean()) / df_raw["CDS_basis_bp"].std(ddof=1)

# ===================== INTERPOLATE (time) =====================
value_cols = ["UK_CDS_bp","US_CDS_bp","UK_10Y_bp","US_10Y_bp"]
for c in value_cols:
    df[f"{c}_was_nan"] = df[c].isna()

df[value_cols] = df[value_cols].interpolate(method="time", limit_direction="both")

# Recompute basis on interpolated data
df["CDS_diff_bp"]  = df["UK_CDS_bp"] - df["US_CDS_bp"]
df["Y10_diff_bp"]  = df["UK_10Y_bp"] - df["US_10Y_bp"]
df["CDS_basis_bp"] = df["CDS_diff_bp"] - df["Y10_diff_bp"]
df["basis_z"]      = (df["CDS_basis_bp"] - df["CDS_basis_bp"].mean()) / df["CDS_basis_bp"].std(ddof=1)

# ===================== SAVE =====================
df_raw.to_csv(OUT_RAW, index_label="Date")
df.to_csv(OUT_INT, index_label="Date")
print(f"\nSaved RAW monthly: {OUT_RAW}")
print(f"Saved INTERPOLATED monthly: {OUT_INT}")

filled_stats = {c: int(df[f"{c}_was_nan"].sum()) for c in value_cols}
print("Months filled by interpolation (per series):", filled_stats)

# ===================== PLOTS =====================
# Chart 1: Basis with median & last-point label
fig, ax = plt.subplots(figsize=(10,4))
df["CDS_basis_bp"].plot(ax=ax, title="UK–US CDS Basis (Monthly, bps) — Interpolated")
add_median_line(ax, df["CDS_basis_bp"], "median")
# label last point
last_x = df.index[-1]
last_y = float(df["CDS_basis_bp"].iloc[-1])
label_last_point(ax, last_x, last_y, f"{last_y:.2f} (last)")
plt.xlabel("")
plt.tight_layout()
plt.show()

# Chart 2: Diff series with their own medians & last-point labels
fig, ax = plt.subplots(figsize=(10,4))
df[["CDS_diff_bp","Y10_diff_bp"]].plot(ax=ax, title="UK–US: CDS Diff vs 10Y Yield Diff (bps) — Interpolated")
# medians
add_median_line(ax, df["CDS_diff_bp"],  "CDS diff median")
add_median_line(ax, df["Y10_diff_bp"],  "10Y diff median")
# label last points
last_cds = float(df["CDS_diff_bp"].iloc[-1])
last_y10 = float(df["Y10_diff_bp"].iloc[-1])
label_last_point(ax, last_x, last_cds, f"{last_cds:.2f} (last CDS)")
label_last_point(ax, last_x, last_y10, f"{last_y10:.2f} (last 10Y)")
plt.xlabel("")
plt.tight_layout()
plt.show()

# ===================== SUMMARY =====================
nz = df.dropna(subset=["UK_CDS_bp","US_CDS_bp","UK_10Y_bp","US_10Y_bp"])
summary = {
    "interp_start": str(nz.index.min().date()) if not nz.empty else None,
    "interp_end":   str(nz.index.max().date()) if not nz.empty else None,
    "n_months_interpolated": int(nz.shape[0]),
    "basis_mean_bp": round(df["CDS_basis_bp"].mean(), 2) if not df["CDS_basis_bp"].dropna().empty else None,
    "basis_std_bp":  round(df["CDS_basis_bp"].std(ddof=1), 2) if not df["CDS_basis_bp"].dropna().empty else None,
    "last_basis_bp": round(df["CDS_basis_bp"].iloc[-1], 2) if not df["CDS_basis_bp"].dropna().empty else None,
    "last_basis_z":  round(df["basis_z"].iloc[-1], 2) if not df["basis_z"].dropna().empty else None,
}
print("Summary:", summary)






```
