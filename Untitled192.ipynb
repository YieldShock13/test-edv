```python
# =========================================
# BRENT–WTI SPREAD + Z-SCORE (with manual today)
# also calculates median spread over full sample
# =========================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

LOOKBACK = 252

# -----------------------
# LOAD DATA
# -----------------------

wti = pd.read_csv("Crude Oil WTI Futures Historical Data (8).csv")
brent = pd.read_csv("Brent Oil Futures Historical Data (1).csv")

for df in [wti, brent]:
    df["Date"] = pd.to_datetime(df["Date"])
    df["Price"] = df["Price"].astype(str).str.replace(",", "").astype(float)
    df.sort_values("Date", inplace=True)

wti = wti[["Date","Price"]].rename(columns={"Price":"WTI"})
brent = brent[["Date","Price"]].rename(columns={"Price":"Brent"})

df = pd.merge(wti, brent, on="Date")
df = df.sort_values("Date").reset_index(drop=True)

# -----------------------
# ADD MANUAL TODAY DATA
# -----------------------

today_row = pd.DataFrame({
    "Date":[pd.Timestamp.today().normalize()],
    "WTI":[79.75],
    "Brent":[84.31]
})

df = pd.concat([df, today_row], ignore_index=True)

# -----------------------
# SPREAD
# -----------------------

df["spread"] = df["Brent"] - df["WTI"]

# -----------------------
# Z SCORE
# -----------------------

df["mean"] = df["spread"].rolling(LOOKBACK).mean()
df["std"] = df["spread"].rolling(LOOKBACK).std()

df["z"] = (df["spread"] - df["mean"]) / df["std"]

# -----------------------
# MEDIAN SPREAD
# -----------------------

median_spread = df["spread"].median()

print("Median spread over sample:", round(median_spread,3))

print("\nLatest spread:", round(df["spread"].iloc[-1],3))
print("Latest Z-score:", round(df["z"].iloc[-1],3))

# -----------------------
# CHART
# -----------------------

plt.figure(figsize=(12,6))
plt.plot(df["Date"], df["spread"])
plt.title("Brent − WTI Spread")
plt.grid(True)

plt.scatter(df["Date"].iloc[-1], df["spread"].iloc[-1])
plt.text(df["Date"].iloc[-1], df["spread"].iloc[-1],
         f"{df['spread'].iloc[-1]:.2f}")

plt.show()

plt.figure(figsize=(12,6))
plt.plot(df["Date"], df["z"])
plt.axhline(2)
plt.axhline(-2)
plt.axhline(0)

plt.scatter(df["Date"].iloc[-1], df["z"].iloc[-1])
plt.text(df["Date"].iloc[-1], df["z"].iloc[-1],
         f"{df['z'].iloc[-1]:.2f}")

plt.title("Spread Z-Score")
plt.grid(True)
plt.show()
```
