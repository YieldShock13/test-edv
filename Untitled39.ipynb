```python
# === 1. Install Packages ===
!pip install tensorflow keras --quiet

# === 2. Imports ===
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense
from fredapi import Fred
import datetime

# === 3. Setup FRED ===
fred = Fred(api_key='eb09beeb04065b310e4dc47320ab0fce')

# === 4. Pull 10-Year NASDAQ Data ===
end = datetime.datetime.today()
start = end - datetime.timedelta(days=365 * 10)
df = fred.get_series('NASDAQCOM', observation_start=start, observation_end=end).to_frame(name='Price')
df.index = pd.to_datetime(df.index)
df = df.asfreq('D').ffill()

# === 5. Scale Data ===
scaler = MinMaxScaler()
scaled = scaler.fit_transform(df[['Price']])
df['Scaled'] = scaled

# === 6. Create Training Sequences (60-day lookback, t+30 prediction) ===
def create_sequences(data, input_len=60, forecast_horizon=30):
    X, y = [], []
    for i in range(input_len, len(data) - forecast_horizon):
        X.append(data[i - input_len:i])
        y.append(data[i + forecast_horizon])
    return np.array(X), np.array(y)

X, y = create_sequences(scaled, input_len=60, forecast_horizon=30)
X = X.reshape((X.shape[0], X.shape[1], 1))

# === 7. Train LSTM ===
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(60, 1)))
model.add(LSTM(units=50))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
model.fit(X, y, epochs=10, batch_size=32, verbose=1)

# === 8. Forecast Next 30 Days ===
lookback_scaled = df['Scaled'].values[-60:]
input_seq = lookback_scaled.reshape((1, 60, 1))
forecast_scaled = []
current_input = input_seq.copy()

for _ in range(30):
    pred = model.predict(current_input, verbose=0)[0, 0]
    forecast_scaled.append(pred)
    current_input = np.append(current_input[:, 1:, :], [[[pred]]], axis=1)

forecast = scaler.inverse_transform(np.array(forecast_scaled).reshape(-1, 1))

# === 9. Build Index for Forecast ===
forecast_start = df.index[-1] + pd.Timedelta(days=1)
forecast_index = pd.date_range(start=forecast_start, periods=30, freq='D')

# === 10. Plot Forecast ===
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(14, 6))

# Plot training data
ax.plot(df.index, df['Price'], color='gray', alpha=0.6, label='Training Data')

# Plot last 60 days used for input
ax.plot(df.index[-60:], df['Price'][-60:], color='white', linewidth=2, label='Lookback Input (60d)')

# Plot forecast
ax.plot(forecast_index, forecast, color='lime', linewidth=2, label='Forecast (Next 30 Days)')

# Label the last real datapoint
ax.annotate(f'{df["Price"].iloc[-1]:,.0f}',
            xy=(df.index[-1], df["Price"].iloc[-1]),
            xytext=(df.index[-1] + pd.Timedelta(days=1), df["Price"].iloc[-1]),
            fontsize=10, color='white',
            arrowprops=dict(facecolor='white', arrowstyle='->'))

# Label the final forecasted value
ax.annotate(f'{forecast[-1,0]:,.0f}',
            xy=(forecast_index[-1], forecast[-1,0]),
            xytext=(forecast_index[-1] + pd.Timedelta(days=1), forecast[-1,0]),
            fontsize=10, color='lime',
            arrowprops=dict(facecolor='lime', arrowstyle='->'))

# Style
ax.set_title("NASDAQ LSTM Forecast — August 2025 Projection", fontsize=14, weight='bold')
ax.set_xlabel("Date")
ax.set_ylabel("Index Level")
ax.legend(loc='upper left')
ax.grid(True, linestyle='--', color='gray', linewidth=0.3)
fig.text(0.5, 0.02, 'Pearl Gray Equity and Research', ha='center', fontsize=10, color='white', alpha=0.9)
plt.tight_layout(rect=[0, 0.04, 1, 0.96])
plt.show()

# === 11. Output Summary Table ===
summary = pd.DataFrame({
    "Metric": ["Last Known Value", "Forecast End Value", "Forecast Start Date", "Forecast End Date"],
    "Value": [f"{df['Price'].iloc[-1]:,.2f}",
              f"{forecast[-1, 0]:,.2f}",
              forecast_index[0].strftime('%Y-%m-%d'),
              forecast_index[-1].strftime('%Y-%m-%d')]
})
print(summary.to_markdown(index=False))










```
