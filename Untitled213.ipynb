```python
# ==============================
# GLD 252-Day Rolling Z-Score
# ==============================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ------------------------------
# Load Data (adjust filename if needed)
# ------------------------------
file_name = "GLD Historical Data (3).csv"

df = pd.read_csv(file_name)

# ------------------------------
# Clean Data
# ------------------------------
# Assume standard Investing.com format
df.columns = [col.strip() for col in df.columns]

# Convert Date
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')

# Clean Price (remove commas if present)
df['Price'] = df['Price'].astype(str).str.replace(',', '')
df['Price'] = df['Price'].astype(float)

# ------------------------------
# Compute 252-day Z-score
# ------------------------------
window = 252

df['rolling_mean'] = df['Price'].rolling(window).mean()
df['rolling_std'] = df['Price'].rolling(window).std()

df['z_score'] = (df['Price'] - df['rolling_mean']) / df['rolling_std']

# Drop NaNs from rolling
df = df.dropna()

# ------------------------------
# Plot (Professional Style)
# ------------------------------
plt.style.use('seaborn-v0_8-darkgrid')

plt.figure(figsize=(14, 6))

plt.plot(df['Date'], df['z_score'], linewidth=2)

# Reference lines
plt.axhline(0, linestyle='--', linewidth=1)
plt.axhline(2, linestyle='--', linewidth=1)
plt.axhline(-2, linestyle='--', linewidth=1)

# Title + labels (clean, professional)
plt.title('GLD 252-Day Rolling Z-Score', fontsize=16, weight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Z-Score', fontsize=12)

# ------------------------------
# Label final data point
# ------------------------------
last_date = df['Date'].iloc[-1]
last_value = df['z_score'].iloc[-1]

plt.scatter(last_date, last_value)
plt.text(last_date, last_value,
         f'{last_value:.2f}',
         fontsize=11,
         verticalalignment='bottom')

plt.tight_layout()
plt.show()

# ------------------------------
# Print latest value
# ------------------------------
print(f"Latest Z-score: {last_value:.4f}")
```
