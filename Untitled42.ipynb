```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Load data
spx = pd.read_csv("S&P 500 Historical Data (1).csv")
vix = pd.read_csv("VIXCLS.csv")

# Format S&P 500 data
spx["Date"] = pd.to_datetime(spx["Date"], format="%m/%d/%Y")
spx["Price"] = spx["Price"].str.replace(",", "").astype(float)
spx = spx.sort_values("Date")

# Format VIX data
vix["Date"] = pd.to_datetime(vix["observation_date"])
vix = vix.rename(columns={"VIXCLS": "VIX"})
vix = vix[vix["VIX"] != "."]
vix["VIX"] = vix["VIX"].astype(float)

# Merge datasets
df = pd.merge(spx[["Date", "Price"]], vix[["Date", "VIX"]], on="Date", how="inner")

# Realized volatility: 30-day rolling std dev of returns, annualized
df["Return"] = df["Price"].pct_change()
df["RealizedVol"] = df["Return"].rolling(30).std() * (252**0.5) * 100
df = df.dropna()

# Plot
plt.figure(figsize=(14, 6))
plt.plot(df["Date"], df["VIX"], label="VIX (Implied)", linewidth=2, color="deepskyblue")
plt.plot(df["Date"], df["RealizedVol"], label="Realized Vol (30D)", linewidth=2, color="orange")

# End-point labels
plt.text(df["Date"].iloc[-1], df["VIX"].iloc[-1], f"{df['VIX'].iloc[-1]:.2f}%",
         fontsize=10, color="deepskyblue", va='bottom', ha='left')
plt.text(df["Date"].iloc[-1], df["RealizedVol"].iloc[-1], f"{df['RealizedVol'].iloc[-1]:.2f}%",
         fontsize=10, color="orange", va='bottom', ha='left')

# Title and labels
plt.title("Implied vs Realized Volatility", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Volatility (%)", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend()

# Watermark
plt.text(df["Date"].iloc[int(len(df)*0.85)],
         df["RealizedVol"].min(),
         "Pearl Gray Equity and Research", fontsize=10,
         color="gray", ha='left', va='bottom', alpha=0.7)

# Format x-axis
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

plt.tight_layout()
plt.show()












```
