```python
# === INSTALL IF NEEDED ===
!pip install fredapi --quiet
!pip install tensorflow scikit-learn --quiet

# === IMPORTS ===
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LassoCV, RidgeCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
from tensorflow.keras.callbacks import EarlyStopping
import statsmodels.formula.api as smf
from statsmodels.stats.anova import anova_lm
from fredapi import Fred
from matplotlib import rcParams

# === STYLE ===
plt.style.use("dark_background")
rcParams['font.family'] = 'DejaVu Sans'
WATERMARK = "Pearl Gray Equity and Research"

# === FRED API ===
fred = Fred(api_key="58dc48b536135c43b951bc7400547b3f")

# === SERIES ===
series_map = {
    "VIX": "VIXCLS",
    "YieldCurveSlope": "T10Y2Y",
    "FedFunds": "FEDFUNDS",
    "OAS": "BAMLH0A0HYM2"
}

# === LOAD DATA ===
data = {name: fred.get_series(code) for name, code in series_map.items()}
df = pd.DataFrame(data).sort_index()
df = df.resample('M').mean()
df = df[df.index <= "2025-06-30"]

# === AR(3) LAGS ===
df['OAS_L1'] = df['OAS'].shift(1)
df['OAS_L2'] = df['OAS'].shift(2)
df['OAS_L3'] = df['OAS'].shift(3)

# === REGRESSION VARIABLES ===
predictors = ['VIX', 'FedFunds', 'YieldCurveSlope', 'OAS_L1', 'OAS_L2', 'OAS_L3']
target = 'OAS'
df_model = df.dropna(subset=predictors + [target])
X = df_model[predictors]
y = df_model[target]

# === SCALE ===
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# === OLS with AR(3) + Robust SE ===
formula = f'{target} ~ ' + ' + '.join(predictors)
ols_model = smf.ols(formula=formula, data=df_model).fit(cov_type='HAC', cov_kwds={'maxlags':1})
anova_table = anova_lm(ols_model, typ=2)

# === LASSO & RIDGE ===
lasso_model = LassoCV(cv=5).fit(X_scaled, y)
lasso_coefs = pd.Series(lasso_model.coef_, index=predictors)

ridge_model = RidgeCV(cv=5).fit(X_scaled, y)
ridge_coefs = pd.Series(ridge_model.coef_, index=predictors)

# === LSTM ===
X_lstm = X_scaled.reshape((X_scaled.shape[0], 1, X_scaled.shape[1]))
y_lstm = y.values

lstm_model = Sequential([
    LSTM(50, activation='relu', input_shape=(1, X_scaled.shape[1])),
    Dense(1)
])
lstm_model.compile(optimizer='adam', loss='mse')
lstm_model.fit(X_lstm, y_lstm, epochs=100, verbose=0, callbacks=[EarlyStopping(patience=10, restore_best_weights=True)])
lstm_preds = lstm_model.predict(X_lstm).flatten()

# === OLS Coefs & T-Stats ===
ols_coefs = ols_model.params[predictors]
ols_tvals = ols_model.tvalues[predictors]

# === Dropped Vars ===
lasso_dropped = lasso_coefs[lasso_coefs.abs() < 1e-5].index.tolist()
ridge_dropped = ridge_coefs[ridge_coefs.abs() < 1e-5].index.tolist()

# === PLOT 1: ACTUAL VS PREDICTED ===
plt.figure(figsize=(12, 6))
plt.plot(df_model.index, y, label="Actual OAS", color="white")
plt.plot(df_model.index, ols_model.fittedvalues, '--', label="OLS", color="deepskyblue")
plt.plot(df_model.index, lasso_model.predict(X_scaled), '--', label="Lasso", color="orange")
plt.plot(df_model.index, ridge_model.predict(X_scaled), '--', label="Ridge", color="lime")
plt.plot(df_model.index, lstm_preds, '--', label="LSTM", color="violet")
for i, (label, series, color) in enumerate([
    ('Actual', y, "white"),
    ('OLS', ols_model.fittedvalues, "deepskyblue"),
    ('Lasso', pd.Series(lasso_model.predict(X_scaled), index=df_model.index), "orange"),
    ('Ridge', pd.Series(ridge_model.predict(X_scaled), index=df_model.index), "lime"),
    ('LSTM', pd.Series(lstm_preds, index=df_model.index), "violet")]):
    plt.annotate(f"{series.iloc[-1]:.2f}", (series.index[-1], series.iloc[-1]), color=color, xytext=(25, -20*i), textcoords='offset points')
plt.title("Actual vs Fitted OAS")
plt.legend()
plt.text(0.01, 0.01, WATERMARK, transform=plt.gca().transAxes, fontsize=8, alpha=0.5)
plt.tight_layout()
plt.show()

# === PLOT 2: T-STATISTICS ===
plt.figure(figsize=(10, 5))
bars = plt.bar(ols_tvals.index, ols_tvals.values, color='deepskyblue')
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f"{bar.get_height():.2f}", ha='center', va='bottom')
plt.axhline(2, color='red', linestyle='--')
plt.axhline(-2, color='red', linestyle='--')
plt.title("OLS T-Statistics (with AR3 terms)")
plt.ylabel("t-value")
plt.text(0.01, 0.01, WATERMARK, transform=plt.gca().transAxes, fontsize=8, alpha=0.5)
plt.tight_layout()
plt.show()

# === PLOT 3: COEFFICIENT COMPARISON ===
coef_df = pd.DataFrame({
    'OLS': ols_coefs,
    'Lasso': lasso_coefs,
    'Ridge': ridge_coefs
})
coef_df.plot(kind='bar', figsize=(10, 5))
plt.axhline(0, color='white', linestyle='--', linewidth=1)
plt.title("Coefficient Comparison (Incl. AR3)")
plt.ylabel("Coefficient")
plt.text(0.01, 0.01, WATERMARK, transform=plt.gca().transAxes, fontsize=8, alpha=0.5)
plt.tight_layout()
plt.show()

# === TEXT OUTPUTS ===
print(f"\n📆 Regression period: {df_model.index[0].strftime('%Y-%m-%d')} to {df_model.index[-1].strftime('%Y-%m-%d')}")
print("\n📄 OLS Summary (with AR3):\n", ols_model.summary())
print("\n🔬 ANOVA Table (type 2):\n", anova_table.round(4))
print("\n🧨 Dropped by Lasso:", lasso_dropped if lasso_dropped else "None")
print("🧨 Dropped by Ridge:", ridge_dropped if ridge_dropped else "None")
print("\n📊 Coefficient Table:\n", coef_df.round(4))




























```
