```python
import pandas as pd
import matplotlib.pyplot as plt

# === STEP 1: Load CSV ===
# Upload manually in Colab before running this code
tp = pd.read_csv('THREEFYTP10 (3).csv')

# === STEP 2: Clean and prepare ===
tp.columns = ['Date', 'TermPremium']
tp['Date'] = pd.to_datetime(tp['Date'])
tp['TermPremium'] = pd.to_numeric(tp['TermPremium'], errors='coerce')
tp.dropna(inplace=True)
latest_tp = tp.iloc[-1]

# === STEP 3: Plot ===
plt.figure(figsize=(14, 6))
plt.plot(tp['Date'], tp['TermPremium'], color='deepskyblue', linewidth=2)
plt.axhline(0, color='white', linewidth=0.8, linestyle='--')

# End label
plt.text(latest_tp['Date'], latest_tp['TermPremium'], f"{latest_tp['TermPremium']:.2f}%",
         color='deepskyblue', fontsize=10, fontname='DejaVu Sans', ha='left', va='center')

# Titles and axis
plt.title('U.S. 10Y Term Premium (ACM Model)', fontsize=18, fontname='DejaVu Sans', weight='bold', color='white')
plt.xlabel('Date', fontsize=12, fontname='DejaVu Sans', color='white')
plt.ylabel('Term Premium (%)', fontsize=12, fontname='DejaVu Sans', color='white')

# Watermark (shifted up)
plt.text(0.99, 0.06, 'Pearl Gray Equity and Research', transform=plt.gca().transAxes,
         fontsize=10, fontname='DejaVu Sans', color='gray', ha='right')

# Style
plt.gca().set_facecolor('black')
plt.gcf().patch.set_facecolor('black')
plt.tick_params(colors='white')
plt.xticks(color='white', fontsize=10, fontname='DejaVu Sans')
plt.yticks(color='white', fontsize=10, fontname='DejaVu Sans')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()




```
