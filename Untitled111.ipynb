```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import os
from math import log, sqrt, exp, erf

# =========================================================
# 1. Helpers
# =========================================================

def load_investing_price(path, price_col="Price"):
    df = pd.read_csv(path)
    df["Date"] = pd.to_datetime(df["Date"])
    df[price_col] = (
        df[price_col]
        .astype(str)
        .str.replace(",", "", regex=False)
        .str.replace("-", "", regex=False)
    )
    df[price_col] = pd.to_numeric(df[price_col], errors="coerce")
    df = df.dropna(subset=[price_col])
    # keep only Date + Price so we don't drag Open/High/Low/Vol/Change%
    df = df[["Date", price_col]]
    df = df.sort_values("Date").set_index("Date")
    # month-end
    out = df.resample("ME").last()
    return out

def parse_yield_table(text_block):
    rows = []
    for raw in text_block.strip().splitlines():
        line = raw.strip()
        if not line:
            continue
        parts = line.split()
        if len(parts) < 5:
            continue
        # last 4 tokens: something like 'Nov','06,','2025','3.83%'
        date_tokens = parts[-4:-1]
        y_token = parts[-1]
        date_str = " ".join(date_tokens[:3]).replace(",", "")
        try:
            dt = pd.to_datetime(date_str)
        except Exception:
            continue
        try:
            y = float(y_token.replace("%", ""))
        except Exception:
            continue
        rows.append((dt, y))
    if not rows:
        raise ValueError("No valid yield rows parsed  check the pasted text.")
    df = pd.DataFrame(rows, columns=["Date", "yield_pct"])
    df = df.set_index("Date").sort_index()
    return df

def bs_call_price(S, K, T, r, sigma):
    if T <= 0 or sigma <= 0 or S <= 0 or K <= 0:
        return 0.0
    try:
        d1 = (log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * sqrt(T))
        d2 = d1 - sigma * sqrt(T)
        N = lambda x: 0.5 * (1.0 + erf(x / sqrt(2.0)))
        return S * N(d1) - K * exp(-r * T) * N(d2)
    except Exception:
        return 0.0

def max_drawdown(equity):
    eq = equity.values.astype(float)
    peak = np.maximum.accumulate(eq)
    dd = eq / peak - 1.0
    return dd.min() * 100.0, pd.Series(dd, index=equity.index)

def var_95(ret):
    if len(ret) == 0:
        return float("nan")
    return np.quantile(ret, 0.05) * 100.0

def beta_vs(port_ret, mkt_ret):
    r, m = port_ret.align(mkt_ret, join="inner")
    if len(r) < 2:
        return float("nan")
    cov = np.cov(r, m)[0, 1]
    var = np.var(m)
    return cov / var if var != 0 else float("nan")

def sharpe_ratio(ret, rf_month):
    r, rf = ret.align(rf_month, join="inner")
    ex = r - rf
    std = ex.std(ddof=0)
    if std == 0:
        return float("nan")
    return (ex.mean() / std) * math.sqrt(12.0)

def treynor_ratio(ret, rf_month, beta):
    if beta == 0 or math.isnan(beta):
        return float("nan")
    r, rf = ret.align(rf_month, join="inner")
    rf_ann = rf.mean() * 12.0
    cagr = (1.0 + r).prod() ** (12.0 / len(r)) - 1.0
    excess_ann = cagr - rf_ann
    return excess_ann / beta

# =========================================================
# 2. Load prices
# =========================================================

base = "/content"  # <- CHANGE if your files sit elsewhere

spx_m = load_investing_price(os.path.join(base, "S&P 500 Historical Data (6).csv")) \
          .rename(columns={"Price": "spx_price"})
vix_m = load_investing_price(os.path.join(base, "CBOE Volatility Index Historical Data (4).csv")) \
          .rename(columns={"Price": "vix_price"})
stx_m = load_investing_price(os.path.join(base, "STXGOVJ Historical Data (1).csv")) \
          .rename(columns={"Price": "stx_price"})
agg_m = load_investing_price(os.path.join(base, "AGG Historical Data.csv")) \
          .rename(columns={"Price": "agg_price"})
us1m_m = load_investing_price(os.path.join(base, "United States 1-Month Bond Yield Historical Data.csv")) \
          .rename(columns={"Price": "rf_yield_pct"})

# =========================================================
# 3. Dividend / yield tables (full text)
# =========================================================

STX_YIELD_TEXT = """
Oct 22, 2025 36.17 B Oct 27, 2025 9.37%
Jul 16, 2025 4.57 B Jul 21, 2025 8.92%
Apr 15, 2025 34.89 B Apr 22, 2025 9.41%
Jan 15, 2025 13.14 B Jan 20, 2025 9.22%
Oct 16, 2024 25.90 B Oct 21, 2024 8.13%
Jul 17, 2024 5.88 B Jul 22, 2024 9.32%
Apr 24, 2024 33.73 B Apr 29, 2024 9.98%
Jan 17, 2024 5.33 B Jan 22, 2024 9.57%
Oct 18, 2023 32.95 B Oct 23, 2023 10.96%
Jul 26, 2023 6.54 B Jul 31, 2023 10.82%
Apr 19, 2023 32.61 B Apr 24, 2023 12.44%
Jan 18, 2023 14.27 B Jan 23, 2023 8.44%
Oct 19, 2022 34.15 B Oct 24, 2022 12.63%
Aug 03, 2022 22.46 B Aug 08, 2022 8.18%
Jan 19, 2022 47.71 B Jan 26, 2022 9.98%
Jul 14, 2021 37.69 B Jul 21, 2021 8.97%
Jan 13, 2021 36.07 B Jan 20, 2021 7.92%
Jul 22, 2020 26.28 B Jul 27, 2020 3.68%
"""

AGG_YIELD_TEXT = """
Nov 03, 2025 0.33 1M Nov 06, 2025 3.83%
Oct 01, 2025 0.33 1M Oct 06, 2025 3.83%
Sep 02, 2025 0.33 1M Sep 05, 2025 3.84%
Aug 01, 2025 0.33 1M Aug 06, 2025 3.86%
Jul 01, 2025 0.32 1M Jul 07, 2025 3.81%
Jun 02, 2025 0.32 1M Jun 05, 2025 3.83%
May 01, 2025 0.32 1M May 06, 2025 3.77%
Apr 01, 2025 0.33 1M Apr 04, 2025 3.75%
Mar 03, 2025 0.30 1M Mar 06, 2025 3.70%
Feb 03, 2025 0.32 1M Feb 06, 2025 3.75%
Dec 18, 2024 0.32 1M Dec 23, 2024 3.70%
Dec 02, 2024 0.31 1M Dec 05, 2024 3.34%
Nov 01, 2024 0.31 1M Nov 06, 2024 3.61%
Oct 01, 2024 0.31 1M Oct 04, 2024 3.48%
Sep 03, 2024 0.31 1M Sep 06, 2024 3.46%
Aug 01, 2024 0.30 1M Aug 06, 2024 3.45%
Jul 01, 2024 0.30 1M Jul 05, 2024 3.48%
Jun 03, 2024 0.30 1M Jun 07, 2024 3.46%
May 01, 2024 0.29 1M May 07, 2024 3.46%
Apr 01, 2024 0.29 1M Apr 05, 2024 3.31%
Mar 01, 2024 0.29 1M Mar 07, 2024 3.30%
Feb 01, 2024 0.29 1M Feb 07, 2024 3.18%
Dec 14, 2023 0.28 1M Dec 20, 2023 3.16%
Dec 01, 2023 0.27 1M Dec 07, 2023 2.93%
Nov 01, 2023 0.28 1M Nov 07, 2023 3.25%
Oct 02, 2023 0.26 1M Oct 06, 2023 3.13%
Sep 01, 2023 0.26 1M Sep 08, 2023 2.99%
Aug 01, 2023 0.26 1M Aug 07, 2023 2.91%
Jul 03, 2023 0.25 1M Jul 10, 2023 2.83%
Jun 01, 2023 0.26 1M Jun 07, 2023 2.75%
May 01, 2023 0.25 1M May 05, 2023 2.64%
Apr 03, 2023 0.25 1M Apr 07, 2023 2.58%
Mar 01, 2023 0.23 1M Mar 07, 2023 2.54%
Feb 01, 2023 0.25 1M Feb 07, 2023 2.40%
Dec 15, 2022 0.23 1M Dec 21, 2022 2.33%
Dec 01, 2022 0.23 1M Dec 07, 2022 2.13%
Nov 01, 2022 0.22 1M Nov 07, 2022 2.26%
Oct 03, 2022 0.21 1M Oct 07, 2022 2.19%
Sep 01, 2022 0.20 1M Sep 08, 2022 2.06%
Aug 01, 2022 0.20 1M Aug 05, 2022 1.96%
Jul 01, 2022 0.19 1M Jul 08, 2022 1.97%
Jun 01, 2022 0.19 1M Jun 07, 2022 1.91%
May 02, 2022 0.18 1M May 06, 2022 1.90%
Apr 01, 2022 0.16 1M Apr 07, 2022 1.82%
Mar 01, 2022 0.16 1M Mar 07, 2022 1.78%
Feb 01, 2022 0.16 1M Feb 07, 2022 1.78%
Dec 16, 2021 0.12 1M Dec 22, 2021 1.77%
Dec 01, 2021 0.16 1M Dec 07, 2021 1.66%
Nov 01, 2021 0.18 1M Nov 05, 2021 1.81%
Oct 01, 2021 0.18 1M Oct 07, 2021 1.82%
Sep 01, 2021 0.17 1M Sep 08, 2021 1.82%
Aug 02, 2021 0.17 1M Aug 06, 2021 1.83%
Jul 01, 2021 0.16 1M Jul 08, 2021 1.89%
Jun 01, 2021 0.17 1M Jun 07, 2021 1.96%
May 03, 2021 0.17 1M May 07, 2021 2.00%
Apr 01, 2021 0.17 1M Apr 08, 2021 2.07%
Mar 01, 2021 0.19 1M Mar 05, 2021 2.11%
Feb 01, 2021 0.19 1M Feb 05, 2021 2.11%
Dec 17, 2020 0.13 1M Dec 23, 2020 2.15%
Dec 01, 2020 0.21 1M Dec 07, 2020 2.03%
Nov 02, 2020 0.19 1M Nov 06, 2020 2.25%
Oct 01, 2020 0.20 1M Oct 07, 2020 2.28%
Sep 01, 2020 0.20 1M Sep 08, 2020 2.32%
Aug 03, 2020 0.21 1M Aug 07, 2020 2.36%
Jul 01, 2020 0.22 1M Jul 08, 2020 2.43%
Jun 01, 2020 0.21 1M Jun 05, 2020 2.46%
May 01, 2020 0.24 1M May 07, 2020 2.52%
Apr 01, 2020 0.25 1M Apr 07, 2020 2.57%
Mar 02, 2020 0.24 1M Mar 06, 2020 2.57%
Feb 03, 2020 0.24 1M Feb 07, 2020 2.63%
Dec 19, 2019 0.19 1M Dec 26, 2019 2.70%
Dec 02, 2019 0.25 1M Dec 06, 2019 2.52%
Nov 01, 2019 0.25 1M Nov 07, 2019 2.71%
Oct 01, 2019 0.26 1M Oct 07, 2019 2.71%
Sep 03, 2019 0.26 1M Sep 09, 2019 2.68%
Aug 01, 2019 0.26 1M Aug 07, 2019 2.72%
Jul 01, 2019 0.25 1M Jul 08, 2019 2.71%
Jun 03, 2019 0.26 1M Jun 07, 2019 2.72%
May 01, 2019 0.26 1M May 07, 2019 2.75%
Apr 01, 2019 0.26 1M Apr 05, 2019 2.72%
Mar 01, 2019 0.27 1M Mar 07, 2019 2.75%
Feb 01, 2019 0.27 1M Feb 07, 2019 2.72%
Dec 18, 2018 0.21 1M Dec 24, 2018 2.73%
Dec 03, 2018 0.25 1M Dec 07, 2018 2.56%
Nov 01, 2018 0.26 1M Nov 07, 2018 2.67%
Oct 01, 2018 0.25 1M Oct 05, 2018 2.62%
Sep 04, 2018 0.23 1M Sep 10, 2018 2.58%
Aug 01, 2018 0.25 1M Aug 07, 2018 2.59%
Jul 02, 2018 0.23 1M Jul 09, 2018 2.56%
Jun 01, 2018 0.25 1M Jun 07, 2018 2.55%
May 01, 2018 0.24 1M May 07, 2018 2.55%
Apr 02, 2018 0.24 1M Apr 06, 2018 2.51%
Mar 01, 2018 0.24 1M Mar 07, 2018 2.51%
Feb 01, 2018 0.24 1M Feb 07, 2018 2.46%
Dec 21, 2017 0.04 1M Dec 28, 2017 2.42%
Dec 14, 2017 0.10 O Dec 22, 2017 2.38%
Dec 01, 2017 0.23 1M Dec 07, 2017 2.52%
Nov 01, 2017 0.23 1M Nov 07, 2017 2.49%
Oct 02, 2017 0.23 1M Oct 06, 2017 2.47%
Sep 01, 2017 0.23 1M Sep 08, 2017 2.43%
Aug 01, 2017 0.23 1M Aug 07, 2017 2.42%
Jul 03, 2017 0.23 O Jul 10, 2017 2.41%
Jun 01, 2017 0.23 1M Jun 07, 2017 2.39%
May 01, 2017 0.24 1M May 05, 2017 2.39%
Apr 03, 2017 0.22 1M Apr 07, 2017 2.37%
Mar 01, 2017 0.22 1M Mar 07, 2017 2.37%
Feb 01, 2017 0.22 1M Feb 07, 2017 2.38%
Dec 22, 2016 0.25 1M Dec 29, 2016 2.40%
Dec 01, 2016 0.20 1M Dec 07, 2016 2.36%
Nov 01, 2016 0.21 1M Nov 07, 2016 2.45%
Oct 03, 2016 0.20 1M Oct 07, 2016 2.43%
Sep 01, 2016 0.21 1M Sep 08, 2016 2.43%
Aug 01, 2016 0.21 1M Aug 05, 2016 2.43%
Jul 01, 2016 0.21 1M Jul 08, 2016 2.44%
Jun 01, 2016 0.22 1M Jun 07, 2016 2.49%
May 02, 2016 0.21 1M May 06, 2016 2.48%
Apr 01, 2016 0.22 1M Apr 07, 2016 2.48%
Mar 01, 2016 0.22 1M Mar 07, 2016 2.48%
Feb 01, 2016 0.23 1M Feb 05, 2016 2.47%
Dec 24, 2015 0.22 1M Dec 31, 2015 2.45%
Dec 01, 2015 0.37 O Dec 07, 2015 2.57%
Nov 02, 2015 0.21 O Nov 06, 2015 2.47%
Oct 01, 2015 0.21 O Oct 07, 2015 2.45%
Sep 01, 2015 0.22 O Sep 08, 2015 2.45%
Aug 03, 2015 0.21 O Aug 07, 2015 2.41%
Jul 01, 2015 0.22 O Jul 08, 2015 2.42%
Jun 01, 2015 0.21 O Jun 05, 2015 2.37%
May 01, 2015 0.21 O May 07, 2015 2.35%
Apr 01, 2015 0.21 O Apr 08, 2015 2.33%
Mar 02, 2015 0.19 O Mar 06, 2015 2.33%
Feb 02, 2015 0.18 O Feb 06, 2015 2.32%
Dec 24, 2014 0.37 O Dec 31, 2014 2.40%
Dec 01, 2014 0.28 O Dec 05, 2014 2.24%
Nov 03, 2014 0.19 O Nov 07, 2014 2.18%
Oct 01, 2014 0.19 O Oct 07, 2014 2.18%
"""

SPY_YIELD_TEXT = """
Sep 19, 2025 1.83 3M Oct 31, 2025 1.10%
Jun 20, 2025 1.76 3M Jul 31, 2025 1.20%
Mar 21, 2025 1.70 3M Apr 30, 2025 1.27%
Dec 20, 2024 1.97 3M Jan 31, 2025 1.21%
Sep 20, 2024 1.75 3M Oct 31, 2024 1.23%
Jun 21, 2024 1.76 3M Jul 31, 2024 1.25%
Mar 15, 2024 1.59 3M Apr 30, 2024 1.31%
Dec 15, 2023 1.91 3M Jan 31, 2024 1.41%
Sep 15, 2023 1.58 3M Oct 31, 2023 1.45%
Jun 16, 2023 1.64 3M Jul 31, 2023 1.47%
Mar 17, 2023 1.51 3M Apr 28, 2023 1.63%
Dec 16, 2022 1.78 3M Jan 31, 2023 1.62%
Sep 16, 2022 1.60 3M Oct 31, 2022 1.58%
Jun 17, 2022 1.58 3M Jul 29, 2022 1.64%
Mar 18, 2022 1.37 3M Apr 29, 2022 1.32%
Dec 17, 2021 1.64 3M Jan 31, 2022 1.23%
Sep 17, 2021 1.43 3M Oct 29, 2021 1.27%
Jun 18, 2021 1.38 3M Jul 30, 2021 1.32%
Mar 19, 2021 1.28 3M Apr 30, 2021 1.42%
Dec 18, 2020 1.58 3M Jan 29, 2021 1.53%
Sep 18, 2020 1.34 3M Oct 30, 2020 1.69%
Jun 19, 2020 1.37 3M Jul 31, 2020 1.84%
Mar 20, 2020 1.41 3M Apr 30, 2020 2.41%
Dec 20, 2019 1.57 3M Jan 31, 2020 1.75%
"""

# Parse yield tables
stx_y_df = parse_yield_table(STX_YIELD_TEXT)
agg_y_df = parse_yield_table(AGG_YIELD_TEXT)
spy_y_df = parse_yield_table(SPY_YIELD_TEXT)

stx_y_ann = stx_y_df["yield_pct"] / 100.0
agg_y_ann = agg_y_df["yield_pct"] / 100.0
spy_y_ann = spy_y_df["yield_pct"] / 100.0

# =========================================================
# 4. Merge monthly series + build total returns
# =========================================================

df = spx_m.join([vix_m, stx_m, agg_m, us1m_m], how="inner")
df = df[["spx_price", "vix_price", "stx_price", "agg_price", "rf_yield_pct"]]

# regression window: Jan 2022 onward
df = df.loc["2022-01-31":]

# align annual yields -> monthly
stx_y_month = stx_y_ann.reindex(df.index, method="ffill").bfill() / 12.0
agg_y_month = agg_y_ann.reindex(df.index, method="ffill").bfill() / 12.0
spy_y_month = spy_y_ann.reindex(df.index, method="ffill").bfill() / 12.0

rf_month = df["rf_yield_pct"] / 100.0 / 12.0

# price returns
spx_ret_p = df["spx_price"].pct_change().fillna(0.0)
stx_ret_p = df["stx_price"].pct_change().fillna(0.0)
agg_ret_p = df["agg_price"].pct_change().fillna(0.0)

# total returns
spx_ret = spx_ret_p + spy_y_month
stx_ret = stx_ret_p + stx_y_month
agg_ret = agg_ret_p + agg_y_month

# =========================================================
# 5. Option overlay on STXGOV collateral
# =========================================================

init_notional = 1_000_000.0
T = 1.0 / 12.0

cols_overlay = ["Overlay_ATM", "Overlay_ITM", "Overlay_OTM"]
equity = pd.DataFrame(
    index=df.index,
    columns=cols_overlay + ["BH_60_40", "BH_SPX", "BetaBench"],
    dtype=float
)

collat_eq = pd.Series(index=df.index, dtype=float)

opt_pnl = pd.DataFrame(index=df.index, columns=["ATM", "ITM", "OTM"], dtype=float)
opt_budget = pd.Series(index=df.index, dtype=float)
participation = pd.DataFrame(index=df.index, columns=["ATM", "ITM", "OTM"], dtype=float)
cost_pct = pd.DataFrame(index=df.index, columns=["ATM", "ITM", "OTM"], dtype=float)

collat_prev = init_notional
eq_60_prev = init_notional
eq_spx_prev = init_notional

for i, (date, row) in enumerate(df.iterrows()):
    S = row["spx_price"]
    vix = row["vix_price"]
    r_stx = stx_ret.loc[date]
    r_spx = spx_ret.loc[date]
    r_agg = agg_ret.loc[date]

    collat_curr = collat_prev * (1.0 + r_stx)
    income = collat_prev * stx_y_month.loc[date]
    opt_budget.loc[date] = max(income, 0.0)

    if i == 0:
        for k in ["ATM", "ITM", "OTM"]:
            opt_pnl.loc[date, k] = 0.0
            participation.loc[date, k] = 0.0
            cost_pct.loc[date, k] = 0.0
        collat_prev = collat_curr
        eq_60_prev = eq_60_prev * (1.0 + 0.6 * r_spx + 0.4 * r_agg)
        eq_spx_prev = eq_spx_prev * (1.0 + r_spx)
        equity.loc[date, "BH_60_40"] = eq_60_prev
        equity.loc[date, "BH_SPX"] = eq_spx_prev
        collat_eq.loc[date] = collat_curr
        continue

    prev_date = df.index[i - 1]
    S_prev = df.loc[prev_date, "spx_price"]
    vix_prev = df.loc[prev_date, "vix_price"]
    rf_prev = df.loc[prev_date, "rf_yield_pct"] / 100.0  # annual decimal

    sigma = max(vix_prev / 100.0, 0.0001)
    r_cont = rf_prev

    K_atm = S_prev
    K_itm = 0.95 * S_prev
    K_otm = 1.05 * S_prev

    C_atm = max(bs_call_price(S_prev, K_atm, T, r_cont, sigma), 0.0)
    C_itm = max(bs_call_price(S_prev, K_itm, T, r_cont, sigma), 0.0)
    C_otm = max(bs_call_price(S_prev, K_otm, T, r_cont, sigma), 0.0)

    budget = opt_budget.loc[date]

    def trade_one(C, K):
        if C <= 0 or budget <= 0:
            return 0.0, 0.0, 0.0
        q = budget / C
        payoff = max(S - K, 0.0) * q
        net = payoff - budget
        part = (q * S_prev) / collat_prev
        return net, part, C / S_prev

    net_atm, part_atm, cpct_atm = trade_one(C_atm, K_atm)
    net_itm, part_itm, cpct_itm = trade_one(C_itm, K_itm)
    net_otm, part_otm, cpct_otm = trade_one(C_otm, K_otm)

    opt_pnl.loc[date, "ATM"] = net_atm
    opt_pnl.loc[date, "ITM"] = net_itm
    opt_pnl.loc[date, "OTM"] = net_otm

    participation.loc[date, "ATM"] = part_atm
    participation.loc[date, "ITM"] = part_itm
    participation.loc[date, "OTM"] = part_otm

    cost_pct.loc[date, "ATM"] = cpct_atm * 100.0
    cost_pct.loc[date, "ITM"] = cpct_itm * 100.0
    cost_pct.loc[date, "OTM"] = cpct_otm * 100.0

    collat_prev = collat_curr

    eq_60_prev = eq_60_prev * (1.0 + 0.6 * r_spx + 0.4 * r_agg)
    eq_spx_prev = eq_spx_prev * (1.0 + r_spx)
    equity.loc[date, "BH_60_40"] = eq_60_prev
    equity.loc[date, "BH_SPX"] = eq_spx_prev
    collat_eq.loc[date] = collat_curr

cum_atm = opt_pnl["ATM"].cumsum()
cum_itm = opt_pnl["ITM"].cumsum()
cum_otm = opt_pnl["OTM"].cumsum()

equity["Overlay_ATM"] = collat_eq + cum_atm
equity["Overlay_ITM"] = collat_eq + cum_itm
equity["Overlay_OTM"] = collat_eq + cum_otm

# =========================================================
# 6. Beta-matched benchmark vs Overlay_ATM
# =========================================================

def eq_to_ret(eq):
    return eq.pct_change().dropna()

ret_spx = spx_ret.loc[equity.index]
ret_atm = eq_to_ret(equity["Overlay_ATM"])
ret_60 = eq_to_ret(equity["BH_60_40"])
ret_spx_eq = eq_to_ret(equity["BH_SPX"])

rf_m = rf_month.reindex(equity.index).ffill().bfill()

beta_atm = beta_vs(ret_atm, ret_spx)

beta_eq = pd.Series(index=equity.index, dtype=float)
beta_eq.iloc[0] = init_notional
for i, date in enumerate(equity.index[1:], start=1):
    r_m = spx_ret.loc[date]
    r_rf = rf_m.loc[date]
    total_ret = beta_atm * r_m + (1.0 - beta_atm) * r_rf
    beta_eq.iloc[i] = beta_eq.iloc[i - 1] * (1.0 + total_ret)
equity["BetaBench"] = beta_eq.values

# =========================================================
# 7. Stats
# =========================================================

stats_rows = []
for name in ["Overlay_ATM", "Overlay_ITM", "Overlay_OTM", "BH_60_40", "BH_SPX", "BetaBench"]:
    eq = equity[name].dropna()
    r = eq.pct_change().dropna()
    n = len(r)
    if n == 0:
        continue
    cagr = (eq.iloc[-1] / eq.iloc[0]) ** (12.0 / n) - 1.0
    vol_ann = r.std(ddof=0) * math.sqrt(12.0)
    maxdd, _dd = max_drawdown(eq)
    var95 = var_95(r)
    beta = beta_vs(r, ret_spx)
    sharpe = sharpe_ratio(r, rf_m)
    treynor = treynor_ratio(r, rf_m, beta)
    win_rate = (r > 0).mean() * 100.0
    stats_rows.append([
        name, cagr * 100.0, vol_ann * 100.0, maxdd, var95, sharpe, beta, treynor, win_rate
    ])

stats = pd.DataFrame(
    stats_rows,
    columns=["Portfolio", "CAGR_%", "Vol_ann_%", "MaxDD_%", "VaR95_month_%",
             "Sharpe", "Beta_vs_SPX", "Treynor", "WinRate_%"]
).set_index("Portfolio")

r_atm_align, r_60_align = ret_atm.align(ret_60, join="inner")
tracking_error_ann = (r_atm_align - r_60_align).std(ddof=0) * math.sqrt(12.0) * 100.0

print("===== Performance Summary ({} to {}) =====".format(equity.index[0].date(), equity.index[-1].date()))
print(stats)
print("\nTracking error Overlay_ATM vs 60/40 (ann %): {:.2f}%".format(tracking_error_ann))

# =========================================================
# 8. Charts
# =========================================================

ocean_colors = ["#1f77b4", "#17becf", "#2ca02c", "#00b5ad", "#1f8a70", "#0b3c5d"]

def add_watermark():
    plt.figtext(0.5, 0.02, "Orca's Capital Strategies", ha="center", fontsize=9, alpha=0.35)

# Equity curves
plt.figure(figsize=(10, 6))
plt.plot(equity.index, equity["Overlay_ATM"], label="Overlay ATM", color=ocean_colors[0])
plt.plot(equity.index, equity["Overlay_ITM"], label="Overlay ITM", color=ocean_colors[1])
plt.plot(equity.index, equity["BH_60_40"], label="BH 60/40", color=ocean_colors[2])
plt.plot(equity.index, equity["BH_SPX"], label="BH SPX", color=ocean_colors[3])
plt.plot(equity.index, equity["BetaBench"], label="BetaBench", color=ocean_colors[4], linestyle="--")
plt.title("Equity Curves")
plt.xlabel("Date")
plt.ylabel("Value ($)")
plt.legend()
plt.grid(alpha=0.3)
add_watermark()
plt.show()

# Drawdowns
plt.figure(figsize=(10, 6))
for i, name in enumerate(["Overlay_ATM", "Overlay_ITM", "BH_60_40", "BH_SPX", "BetaBench"]):
    _, dd = max_drawdown(equity[name].dropna())
    plt.plot(dd.index, dd * 100.0, label=name, color=ocean_colors[i])
plt.title("Drawdowns (%)")
plt.xlabel("Date")
plt.ylabel("Drawdown (%)")
plt.legend()
plt.grid(alpha=0.3)
add_watermark()
plt.show()

# Monthly option P&L
plt.figure(figsize=(10, 5))
for i, k in enumerate(["ATM", "ITM", "OTM"]):
    plt.plot(opt_pnl.index, opt_pnl[k], label=k, color=ocean_colors[i])
plt.axhline(0.0, color="black", linewidth=0.8)
plt.title("Monthly Net Option P&L")
plt.xlabel("Date")
plt.ylabel("P&L ($)")
plt.legend()
plt.grid(alpha=0.3)
add_watermark()
plt.show()

# Premium budget
plt.figure(figsize=(10, 4))
plt.plot(opt_budget.index, opt_budget.values, label="Premium budget", color=ocean_colors[0])
plt.title("Monthly Premium Budget (STXGOV income)")
plt.xlabel("Date")
plt.ylabel("Budget ($)")
plt.grid(alpha=0.3)
add_watermark()
plt.show()

# Participation
plt.figure(figsize=(10, 5))
for i, k in enumerate(["ATM", "ITM", "OTM"]):
    plt.plot(participation.index, 100.0 * participation[k], label=k, color=ocean_colors[i])
plt.title("Approx Participation in SPX via Calls (% of collateral)")
plt.xlabel("Date")
plt.ylabel("Participation (%)")
plt.legend()
plt.grid(alpha=0.3)
add_watermark()
plt.show()

# Call cost as % of spot
plt.figure(figsize=(10, 5))
for i, k in enumerate(["ATM", "ITM", "OTM"]):
    plt.plot(cost_pct.index, cost_pct[k], label=k, color=ocean_colors[i])
plt.title("Call Premium as % of SPX Spot (model)")
plt.xlabel("Date")
plt.ylabel("Premium (% of spot)")
plt.legend()
plt.grid(alpha=0.3)
add_watermark()
plt.show()

# Rolling 12M tracking error ATM vs 60/40
te_series = (r_atm_align - r_60_align).rolling(12).std(ddof=0) * math.sqrt(12.0) * 100.0
plt.figure(figsize=(10, 4))
plt.plot(te_series.index, te_series, color=ocean_colors[0])
plt.title("Rolling 12M Tracking Error: Overlay ATM vs 60/40")
plt.xlabel("Date")
plt.ylabel("Tracking Error (ann %, rolling 12m)")
plt.grid(alpha=0.3)
add_watermark()
plt.show()










```
