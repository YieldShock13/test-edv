```python
# === Mortgage  10Y Spread (Quarterly by ACTUAL last date) + Z-Score ===
# Assumes these files are already in your Colab working directory:
#   "MORTGAGE30US (3).csv"   (FRED: MORTGAGE30US)
#   "DGS10 (9).csv"          (FRED: DGS10)
#
# Requirements:
# - No seaborn; one figure per chart.
# - Font must exist in Colab: DejaVu Sans.
# - Watermark: "Orca's Capital Strategies"
# - Cutoff date: use 2025-09-08 (start of September), not quarter-end.
# - Quarterly sampling uses the ACTUAL last available observation in each quarter
#   and preserves its true calendar date (e.g., 2025-09-08), not 2025-09-30.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import findfont, FontProperties

# ---------------- CONFIG ----------------
MORT_PATH   = "MORTGAGE30US (3).csv"
UST10_PATH  = "DGS10 (9).csv"
CUTOFF_DATE = pd.Timestamp("2025-09-08")   # << your requested cutoff

TITLE_PREFIX = "Mortgage Rate  10Y Treasury"
WATERMARK    = "Orca's Capital Strategies"

# Matplotlib font (bundled in Colab)
plt.rcParams["font.family"] = "DejaVu Sans"
fp = FontProperties(fname=findfont("DejaVu Sans"))

# ---------------- HELPERS ----------------
def read_fred_series(path: str) -> pd.Series:
    """Read a FRED CSV (columns: observation_date + value), return datetime-indexed Series."""
    df = pd.read_csv(path)
    date_col = next((c for c in df.columns if c.strip().lower() in ("date","observation_date")), None)
    if date_col is None:
        raise ValueError(f"No DATE/observation_date column in {path}")
    val_col = next((c for c in df.columns if c != date_col), None)
    if val_col is None:
        raise ValueError(f"No value column in {path}")
    s = pd.to_numeric(df[val_col].replace(".", np.nan), errors="coerce")
    return pd.Series(s.values, index=pd.to_datetime(df[date_col]), name=val_col).sort_index()

def quarter_last_actual_with_actual_dates(s: pd.Series) -> pd.Series:
    """
    Build a quarterly series where each point is the last ACTUAL date within that quarter,
    preserving the real timestamp (not the quarter-end label).
    """
    s = s.dropna()
    # Restrict to cutoff BEFORE building quarters
    s = s.loc[:CUTOFF_DATE]
    # For each quarter, pick the max (last) actual date
    last_dates = s.index.to_series().groupby(s.index.to_period("Q")).max()
    out = s.reindex(last_dates.values)
    out.index = last_dates.values  # keep the TRUE dates
    return out.sort_index()

def annotate_last(ax, s: pd.Series, fmt="{:.0f}", units="bps"):
    if s.empty:
        return
    last_x, last_y = s.index[-1], s.iloc[-1]
    ax.annotate(f"{fmt.format(last_y)} {units} ({last_x.strftime('%Y-%m-%d')})",
                xy=(last_x, last_y), xytext=(10, 0), textcoords="offset points",
                ha="left", va="center", fontproperties=fp)

def add_watermark(ax, text=WATERMARK):
    ax.text(0.995, -0.12, text, transform=ax.transAxes,
            ha="right", va="top", alpha=0.5, fontsize=10, fontproperties=fp)

# ---------------- LOAD & PREP ----------------
mort  = read_fred_series(MORT_PATH).rename("mort")   # %
ust10 = read_fred_series(UST10_PATH).rename("ust10") # %

# Align daily series, ffill small gaps, cut at CUTOFF_DATE
daily = pd.concat([mort, ust10], axis=1).sort_index().ffill().dropna()
daily = daily.loc[:CUTOFF_DATE]

# Spread in basis points
daily["spread_bps"] = (daily["mort"] - daily["ust10"]) * 100.0

# Quarterly series (actual last dates)
q_spread = quarter_last_actual_with_actual_dates(daily["spread_bps"])

# ---------------- CHART 1: Spread ----------------
fig1 = plt.figure(figsize=(12, 5))
ax1 = fig1.gca()
ax1.plot(q_spread.index, q_spread.values)
ax1.set_title(f"{TITLE_PREFIX} Spread (Quarterly, labeled at last ACTUAL date)", fontproperties=fp)
ax1.set_xlabel("Date", fontproperties=fp)
ax1.set_ylabel("Basis Points", fontproperties=fp)
ax1.axhline(0, linestyle="--", linewidth=1)
annotate_last(ax1, q_spread, fmt="{:.0f}", units="bps")
add_watermark(ax1)
plt.show()

# ---------------- CHART 2: Z-Score ----------------
# No smoothing. Z across the available history up to the cutoff.
z = (q_spread - q_spread.mean()) / q_spread.std(ddof=0)

fig2 = plt.figure(figsize=(12, 5))
ax2 = fig2.gca()
ax2.plot(z.index, z.values)
ax2.set_title(f"{TITLE_PREFIX} Spread  Z-Score (Quarterly, labeled at last ACTUAL date)", fontproperties=fp)
ax2.set_xlabel("Date", fontproperties=fp)
ax2.set_ylabel("Z", fontproperties=fp)
for lvl in [0, 1, -1, 2, -2]:
    ax2.axhline(lvl, linestyle=":", linewidth=1)
ax2.annotate(f"{z.iloc[-1]:.2f}  ({z.index[-1].strftime('%Y-%m-%d')})",
             xy=(z.index[-1], z.iloc[-1]), xytext=(10, 0),
             textcoords="offset points", ha="left", va="center", fontproperties=fp)
add_watermark(ax2)
plt.show()

print("Last actual date used:", q_spread.index[-1].date(), "| Spread:", f"{q_spread.iloc[-1]:.0f} bps", "| Z:", f"{z.iloc[-1]:.2f} ")


```
