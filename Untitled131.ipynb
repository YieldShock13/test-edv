```python
# --- Elite diagnostics: Gold/Silver ratio regime evolution (Monthly) ---
# Inputs: XAU_USD, XAG_USD monthly price CSVs (Investing-style)
# Outputs (rolling): Ratio, 12M z-score, ACF, Variance Ratio, Entropy, Half-life, Hurst (supporting)
# Watermark: "Orca's Capital Strategies"
# NOTE: Uses monthly log ratio + monthly log returns (core), rolling windows 36/60 by default.

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager

# ---------------- paths ----------------
GOLD_PATH   = "/mnt/data/XAU_USD Historical Data (6).csv"
SILVER_PATH = "/mnt/data/XAG_USD Historical Data (1).csv"

def _resolve(p):
    if os.path.exists(p): return p
    alt = os.path.basename(p)
    if os.path.exists(alt): return alt
    raise FileNotFoundError(f"File not found: {p} (or {alt})")

GOLD_PATH   = _resolve(GOLD_PATH)
SILVER_PATH = _resolve(SILVER_PATH)

# ---------------- font (oceanic, non-default) ----------------
!wget -q -O /tmp/Pacifico-Regular.ttf https://github.com/google/fonts/raw/main/ofl/pacifico/Pacifico-Regular.ttf
font_manager.fontManager.addfont("/tmp/Pacifico-Regular.ttf")
plt.rcParams["font.family"] = font_manager.FontProperties(fname="/tmp/Pacifico-Regular.ttf").get_name()

# ---------------- style ----------------
WATERMARK = "Orca's Capital Strategies"

bg     = "#041521"
grid   = "#10364a"
text   = "#d7f5ff"
accent = "#ff7a90"
c1     = "#2ee6d6"  # aqua
c2     = "#f7c873"  # gold-ish
c3     = "#cfd8dc"  # silver-grey
c4     = "#8b5cf6"  # violet
c5     = "#22c55e"  # green

def add_watermark(ax, fontsize=28):
    ax.text(
        0.5, 0.5, WATERMARK,
        transform=ax.transAxes, fontsize=fontsize,
        color=text, alpha=0.10, ha="center", va="center", rotation=18
    )

def label_last(ax, x, y, fmt="{:.2f}"):
    s = pd.Series(y).dropna()
    if s.empty: return
    idx = s.index[-1]
    ax.scatter(x.iloc[idx], y.iloc[idx], s=60, color=accent, zorder=6)
    ax.annotate(
        fmt.format(float(y.iloc[idx])),
        (x.iloc[idx], y.iloc[idx]),
        textcoords="offset points", xytext=(10,8),
        color=text, fontsize=12,
        bbox=dict(boxstyle="round,pad=0.3", fc=bg, ec=accent, lw=1.2, alpha=0.95)
    )

def stylize(ax, title, xlabel="Date", ylabel=""):
    ax.set_facecolor(bg)
    ax.set_title(title, fontsize=18, color=text, pad=14)
    ax.set_xlabel(xlabel, color=text)
    ax.set_ylabel(ylabel, color=text)
    ax.grid(True, color=grid, alpha=0.55, linewidth=0.8)
    ax.tick_params(colors=text)
    for s in ax.spines.values():
        s.set_color(grid)

# ---------------- loaders ----------------
def load_price(path, label):
    df = pd.read_csv(path)
    df.columns = [c.strip() for c in df.columns]
    if "Date" not in df.columns or "Price" not in df.columns:
        raise ValueError(f"{label}: expected columns ['Date','Price']; got {list(df.columns)}")
    out = df[["Date","Price"]].copy()
    out["Date"] = pd.to_datetime(out["Date"], errors="coerce")
    out["Price"] = pd.to_numeric(out["Price"].astype(str).str.replace(",","",regex=False), errors="coerce")
    out = out.dropna().sort_values("Date")
    out = out.rename(columns={"Price": f"{label}_price"})
    return out

gold   = load_price(GOLD_PATH, "Gold")
silver = load_price(SILVER_PATH, "Silver")

df = (
    pd.merge(gold, silver, on="Date", how="inner")
      .sort_values("Date")
      .reset_index(drop=True)
)

# ---------------- core series ----------------
# Renowned Gold/Silver ratio = Gold price / Silver price
df["gs_ratio"] = df["Gold_price"] / df["Silver_price"]
df["log_ratio"] = np.log(df["gs_ratio"])
df["r"] = df["log_ratio"].diff()   # monthly log return of the ratio

# 12M z-score on level (trade-style)
z_win = 12
mu = df["gs_ratio"].rolling(z_win).mean()
sd = df["gs_ratio"].rolling(z_win).std(ddof=0)
df["z12"] = (df["gs_ratio"] - mu) / sd

# ---------------- diagnostics: rolling windows ----------------
# Choose windows appropriate for monthly data
ROLL = 60   # main regime window (months); 60 = 5y
ROLL_ALT = 36  # optional shorter lens (months)

# ---- Rolling ACF on returns ----
def rolling_acf1(x, w, lag=1):
    x = pd.Series(x)
    out = pd.Series(index=x.index, dtype=float)
    for i in range(w-1, len(x)):
        a = x.iloc[i-w+1:i+1].dropna()
        if len(a) < w:
            out.iloc[i] = np.nan
            continue
        out.iloc[i] = a.autocorr(lag=lag)
    return out

df["acf1_60"]  = rolling_acf1(df["r"], ROLL, lag=1)
df["acf3_60"]  = rolling_acf1(df["r"], ROLL, lag=3)
df["acf12_60"] = rolling_acf1(df["r"], ROLL, lag=12)

# ---- Rolling Variance Ratio (Lo-MacKinlay style; diagnostics) ----
# VR(k) = Var(sum_{j=0..k-1} r_{t-j}) / (k * Var(r))
def rolling_variance_ratio(r, w, k):
    r = pd.Series(r)
    out = pd.Series(index=r.index, dtype=float)
    for i in range(w-1, len(r)):
        x = r.iloc[i-w+1:i+1].dropna()
        if len(x) < w:
            out.iloc[i] = np.nan
            continue
        rk = x.rolling(k).sum().dropna()
        if rk.empty:
            out.iloc[i] = np.nan
            continue
        vr = rk.var(ddof=0) / (k * x.var(ddof=0))
        out.iloc[i] = vr
    return out

for k in [3, 6, 12]:
    df[f"vr{k}_60"] = rolling_variance_ratio(df["r"], ROLL, k)

# ---- Rolling entropy (quantile-binned returns) ----
# Normalized Shannon entropy in [0,1] using 5 bins
def rolling_entropy_quantiles(x, w, bins=5):
    x = pd.Series(x)
    out = pd.Series(index=x.index, dtype=float)
    for i in range(w-1, len(x)):
        a = x.iloc[i-w+1:i+1].dropna()
        if len(a) < w:
            out.iloc[i] = np.nan
            continue
        # quantile bin edges inside the window
        try:
            qs = np.quantile(a, np.linspace(0, 1, bins+1))
            # ensure strictly increasing edges
            qs = np.unique(qs)
            if len(qs) < 3:
                out.iloc[i] = np.nan
                continue
            b = np.digitize(a, qs[1:-1], right=True)
            counts = np.bincount(b, minlength=bins).astype(float)
            p = counts / counts.sum()
            p = p[p > 0]
            H = -(p * np.log(p)).sum()
            Hn = H / np.log(bins)  # normalize
            out.iloc[i] = Hn
        except Exception:
            out.iloc[i] = np.nan
    return out

df["entropy5_60"] = rolling_entropy_quantiles(df["r"], ROLL, bins=5)

# ---- Rolling half-life of mean reversion on z-score (trade relevance) ----
# AR(1): z_t = a + b z_{t-1} + e; half-life = -ln(2)/ln(b) for 0<b<1
def rolling_half_life(x, w):
    x = pd.Series(x)
    out = pd.Series(index=x.index, dtype=float)
    for i in range(w, len(x)):
        y = x.iloc[i-w+1:i+1].dropna()
        if len(y) < w:
            out.iloc[i] = np.nan
            continue
        y1 = y.iloc[1:].values
        x1 = y.iloc[:-1].values
        X = np.column_stack([np.ones(len(x1)), x1])
        # OLS
        beta = np.linalg.lstsq(X, y1, rcond=None)[0]
        b = float(beta[1])
        if b <= 0 or b >= 1:
            out.iloc[i] = np.nan
        else:
            out.iloc[i] = -np.log(2) / np.log(b)
    return out

df["halflife_60"] = rolling_half_life(df["z12"], ROLL)

# ---- Rolling Hurst (supporting color; simple log-log slope of lagged std) ----
# H approx slope of log(std(diff_lag)) vs log(lag); returns series
def hurst_simple(series, lags=(2,3,4,6,8,12)):
    s = np.asarray(series, dtype=float)
    s = s[np.isfinite(s)]
    if len(s) < max(lags) + 5:
        return np.nan
    tau = []
    for lag in lags:
        diff = s[lag:] - s[:-lag]
        tau.append(np.std(diff, ddof=0))
    tau = np.array(tau)
    if np.any(tau <= 0):
        return np.nan
    x = np.log(np.array(lags))
    y = np.log(tau)
    b = np.polyfit(x, y, 1)[0]
    return float(b)

def rolling_hurst(x, w, lags=(2,3,4,6,8,12)):
    x = pd.Series(x)
    out = pd.Series(index=x.index, dtype=float)
    for i in range(w-1, len(x)):
        a = x.iloc[i-w+1:i+1].dropna()
        if len(a) < w:
            out.iloc[i] = np.nan
            continue
        out.iloc[i] = hurst_simple(a.values, lags=lags)
    return out

df["hurst_60"] = rolling_hurst(df["r"], ROLL)

# ---------------- charts ----------------
plt.figure(figsize=(14,6), facecolor=bg)
ax = plt.gca()
ax.plot(df["Date"], df["gs_ratio"], lw=2.6, color=c2)
add_watermark(ax)
label_last(ax, df["Date"], df["gs_ratio"], fmt="{:.2f}")
stylize(ax, "Gold / Silver Ratio (Monthly)", ylabel="Ratio")
plt.tight_layout(); plt.show()

plt.figure(figsize=(14,6), facecolor=bg)
ax = plt.gca()
ax.plot(df["Date"], df["z12"], lw=2.6, color=c3)
ax.axhline(0, color=grid, lw=1.2)
ax.axhline(2, color=grid, lw=1.0, ls="--")
ax.axhline(-2, color=grid, lw=1.0, ls="--")
add_watermark(ax)
label_last(ax, df["Date"], df["z12"], fmt="{:.2f}")
stylize(ax, "Gold / Silver Ratio  12M Rolling Z-Score", ylabel="Z-score")
plt.tight_layout(); plt.show()

plt.figure(figsize=(14,6), facecolor=bg)
ax = plt.gca()
ax.plot(df["Date"], df["acf1_60"],  lw=2.2, color=c1, label="ACF(1) 60M")
ax.plot(df["Date"], df["acf3_60"],  lw=2.2, color=c4, label="ACF(3) 60M")
ax.plot(df["Date"], df["acf12_60"], lw=2.2, color=c5, label="ACF(12) 60M")
ax.axhline(0, color=grid, lw=1.2)
add_watermark(ax)
label_last(ax, df["Date"], df["acf1_60"], fmt="{:.2f}")
stylize(ax, "Rolling Autocorrelation of Monthly Log Returns (60M window)", ylabel="ACF")
leg = ax.legend(frameon=True)
leg.get_frame().set_facecolor(bg); leg.get_frame().set_edgecolor(grid)
for t in leg.get_texts(): t.set_color(text)
plt.tight_layout(); plt.show()

plt.figure(figsize=(14,6), facecolor=bg)
ax = plt.gca()
ax.plot(df["Date"], df["vr3_60"],  lw=2.2, color=c1, label="VR(3)")
ax.plot(df["Date"], df["vr6_60"],  lw=2.2, color=c4, label="VR(6)")
ax.plot(df["Date"], df["vr12_60"], lw=2.2, color=c5, label="VR(12)")
ax.axhline(1, color=grid, lw=1.2)  # VR=1 random walk
add_watermark(ax)
label_last(ax, df["Date"], df["vr12_60"], fmt="{:.2f}")
stylize(ax, "Rolling Variance Ratio on Monthly Log Returns (60M window)", ylabel="VR(k)")
leg = ax.legend(frameon=True)
leg.get_frame().set_facecolor(bg); leg.get_frame().set_edgecolor(grid)
for t in leg.get_texts(): t.set_color(text)
plt.tight_layout(); plt.show()

plt.figure(figsize=(14,6), facecolor=bg)
ax = plt.gca()
ax.plot(df["Date"], df["entropy5_60"], lw=2.6, color=c1)
ax.set_ylim(0, 1.05)
add_watermark(ax)
label_last(ax, df["Date"], df["entropy5_60"], fmt="{:.2f}")
stylize(ax, "Rolling Normalized Entropy (5-bin) of Monthly Log Returns (60M window)", ylabel="Entropy (01)")
plt.tight_layout(); plt.show()

plt.figure(figsize=(14,6), facecolor=bg)
ax = plt.gca()
ax.plot(df["Date"], df["halflife_60"], lw=2.6, color=c4)
add_watermark(ax)
label_last(ax, df["Date"], df["halflife_60"], fmt="{:.1f}")
stylize(ax, "Rolling Mean-Reversion Half-Life on Z-Score (60M window)", ylabel="Half-life (months)")
plt.tight_layout(); plt.show()

plt.figure(figsize=(14,6), facecolor=bg)
ax = plt.gca()
ax.plot(df["Date"], df["hurst_60"], lw=2.6, color=c5)
ax.axhline(0.5, color=grid, lw=1.2)  # ~random-walk boundary for this proxy
add_watermark(ax)
label_last(ax, df["Date"], df["hurst_60"], fmt="{:.2f}")
stylize(ax, "Rolling Hurst (supporting) on Monthly Log Returns (60M window)", ylabel="H (proxy)")
plt.tight_layout(); plt.show()

# ---------------- quick printout (latest) ----------------
last_row = df.dropna(subset=["r"]).iloc[-1]
print("Latest (raw):")
print(f"Date: {last_row['Date'].date()} | G/S: {last_row['gs_ratio']:.4f} | z12: {df['z12'].iloc[-1]:.4f}")

latest_diag = df.dropna(subset=["vr12_60","acf1_60","entropy5_60","halflife_60","hurst_60"]).iloc[-1]
print("\nLatest (diagnostics, 60M window):")
print(f"VR(3,6,12): {latest_diag['vr3_60']:.3f}, {latest_diag['vr6_60']:.3f}, {latest_diag['vr12_60']:.3f}")
print(f"ACF(1,3,12): {latest_diag['acf1_60']:.3f}, {latest_diag['acf3_60']:.3f}, {latest_diag['acf12_60']:.3f}")
print(f"Entropy(5-bin): {latest_diag['entropy5_60']:.3f} (0=noise-free structure, 1=max randomness)")
print(f"Half-life(z): {latest_diag['halflife_60']:.2f} months")
print(f"Hurst proxy: {latest_diag['hurst_60']:.3f}")

# Optional export
# df.to_csv("gs_ratio_elite_diagnostics_monthly.csv", index=False)

```
