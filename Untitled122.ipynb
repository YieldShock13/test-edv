```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------
# GLOBAL STYLE (aquatic, professional)
# ----------------------------
plt.rcParams.update({
    "font.family": "DejaVu Sans",
    "font.size": 11,
    "axes.titleweight": "bold",
    "axes.labelweight": "regular",
    "axes.edgecolor": "#0b3142",
    "axes.facecolor": "#f4fbff",
    "figure.facecolor": "#f4fbff",
    "axes.grid": True,
    "grid.alpha": 0.25,
    "grid.linestyle": "--",
})

# ----------------------------
# 1. HELPERS
# ----------------------------

def load_price_return(path, name):
    df = pd.read_csv(path)
    df["Date"] = pd.to_datetime(df["Date"])
    df = df.sort_values("Date").set_index("Date")
    df[f"{name}_ret"] = df["Price"].pct_change()
    return df[[f"{name}_ret"]]

def align_to_master(df, master_index):
    return df.reindex(master_index, method="nearest")

# ----------------------------
# 2. LOAD ALL SERIES
# ----------------------------

spy  = load_price_return("SPY Historical Data (6).csv", "SPY")
xli  = load_price_return("XLI Historical Data (1).csv", "XLI")
xlu  = load_price_return("XLU Historical Data (1).csv", "XLU")
gld  = load_price_return("GLD Historical Data.csv", "GLD")
jnk  = load_price_return("JNK Historical Data.csv", "JNK")
vix  = load_price_return("CBOE Volatility Index Historical Data (7).csv", "VIX")

slope = pd.read_csv("T10Y3M (4).csv")
slope["observation_date"] = pd.to_datetime(slope["observation_date"])
slope = slope.sort_values("observation_date").set_index("observation_date")

master_index = spy.index

xli = align_to_master(xli, master_index)
xlu = align_to_master(xlu, master_index)
gld = align_to_master(gld, master_index)
jnk = align_to_master(jnk, master_index)
vix = align_to_master(vix, master_index)

slope = align_to_master(slope, master_index)
slope["slope_chg"] = slope["T10Y3M"].diff()

# ----------------------------
# 3. MERGE & CLEAN
# ----------------------------

data = pd.concat([
    spy, xli, xlu, gld, jnk, vix,
    slope[["slope_chg"]],
], axis=1)

data = data.dropna()

print("\nCOMMON SAMPLE WINDOW:")
print("Start:", data.index.min().date(), " End:", data.index.max().date())
print("Shape:", data.shape)

# ----------------------------
# 4. Z-SCORE INPUTS
# ----------------------------

Z = (data - data.mean()) / data.std()
abn = (Z.abs() > 1.96)

print("\n=== RAW CHANGE SUMMARY ===")
print(data.describe())

print("\n=== Z-SCORE SUMMARY ===")
print(Z.describe())

print("\n=== ABNORMAL COUNTS (|z|>1.96) ===")
print(abn.sum())

# ----------------------------
# 5. RISK-ON INDEX
# ----------------------------

offense = Z[["SPY_ret", "XLI_ret", "JNK_ret", "slope_chg"]].mean(axis=1)
defense = Z[["VIX_ret", "GLD_ret", "XLU_ret"]].mean(axis=1)

risk_on_raw = offense - defense

rmin, rmax = risk_on_raw.min(), risk_on_raw.max()
risk_on_scaled = 6 * (risk_on_raw - rmin) / (rmax - rmin) - 3

low_q, high_q = np.quantile(risk_on_scaled, [0.025, 0.975])
risk_on_win = risk_on_scaled.clip(low_q, high_q)

risk_on_ema = risk_on_scaled.ewm(span=5, adjust=False).mean()
risk_on_win_ema = risk_on_win.ewm(span=5, adjust=False).mean()

risk_panel = pd.DataFrame({
    "RiskOn_raw": risk_on_raw,
    "RiskOn_scaled": risk_on_scaled,
    "RiskOn_win": risk_on_win,
    "RiskOn_ema": risk_on_ema,
    "RiskOn_win_ema": risk_on_win_ema,
})

print("\n=== RISK-ON INDEX SUMMARY ===")
print(risk_panel.describe())

# ----------------------------
# 6. CHART WITH ZONES + WATERMARK
# ----------------------------

p = risk_panel.dropna(subset=["RiskOn_ema"])
fig, ax = plt.subplots(figsize=(15, 6))

# ====== BACKGROUND ZONES ======
ax.axhspan(0.30, 1.5, facecolor="#c9f2ff", alpha=0.30, label="Risk-On Zone")
ax.axhspan(-0.30, 0.30, facecolor="#eaf7ff", alpha=0.55, label="Neutral Zone")
ax.axhspan(-1.5, -0.30, facecolor="#ffe2e2", alpha=0.30, label="Risk-Off Zone")

# 1 band
mu = p["RiskOn_ema"].mean()
sd = p["RiskOn_ema"].std()
ax.fill_between(
    p.index,
    mu - sd, mu + sd,
    alpha=0.22,
    color="#7fd3ff",
    label="1 band"
)

# Main EMA lines
ax.plot(
    p.index, p["RiskOn_ema"],
    linewidth=2.0,
    color="#004c6d",
    label="Risk-On Index (EMA 5)"
)

ax.plot(
    p.index, p["RiskOn_win_ema"],
    linewidth=1.5,
    linestyle="--",
    color="#00b3a4",
    label="Risk-On Index (Winsorised EMA)"
)

# Zero line
ax.axhline(0.0, linestyle="--", linewidth=0.9, color="#4b4b4b")

# Fix invisible ticks
ax.tick_params(axis="both", colors="#0b3142")
ax.xaxis.label.set_color("#0b3142")
ax.yaxis.label.set_color("#0b3142")
ax.title.set_color("#0b3142")

ax.set_title("Composite Risk-On Index (5-Week EMA, Scaled)", pad=18)
ax.set_ylabel("Risk-On Score")
ax.set_xlabel("Date")

ax.set_ylim(-1.5, 1.5)

fig.autofmt_xdate()

# Legend
leg = ax.legend(frameon=True)
leg.get_frame().set_edgecolor("#0b3142")
leg.get_frame().set_alpha(0.92)

# WATERMARK
fig.text(
    0.995, 0.02,
    "Orca's Capital Strategies",
    ha='right',
    va='bottom',
    fontsize=10,
    color="#0b3142",
    alpha=0.30
)

plt.tight_layout()
plt.show()

# ----------------------------
# 7. LAST POINT
# ----------------------------
last = p.iloc[-1]
print(
    f"\nLAST POINT [{p.index[-1].date()}]  "
    f"RiskOn_ema={last['RiskOn_ema']:.2f},  "
    f"RiskOn_win_ema={last['RiskOn_win_ema']:.2f}"
)

















```
