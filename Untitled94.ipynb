```python
# ============================================================
# Mag7 backtest using MONTHLY "Change %" -> QUARTERLY returns (geometric)
# QUARTERLY rebalancing; 12-quarter (36-month) training window on QUARTERLY returns
#
# Portfolios:
#   1) Equal-weight (Quarterly)
#   2) Current market-cap (Quarterly)
#   3) Sharpe-Optimized (bounds 5%–30%, Quarterly regression)
#   4) ER/MVaR Score-Optimized (Δ-normal MVaR from w_prev) @ 95% and @ 97.5%
#   5) ER/Standalone VaR Score-Optimized (empirical per-asset VaR) @ 95% and @ 97.5%
#
# Rules:
# - Source returns are MONTHLY % changes (from CSV). Do not transform them; they are returns.
# - Build QUARTERLY returns by geometric compounding of the three monthly returns in each quarter.
# - The lookback (regression stats) is computed on QUARTERLY returns (12 quarters).
# - Risk-free (DTB3) is an ANNUAL YIELD LEVEL (percent). Use as a level:
#     * For training excess QUARTERLY returns: rf_per_quarter = (annual_level_decimal) / 4
#     * For reporting Sharpe (annual): (annualized mean) - (annual level) over (annualized vol)
# - Rebalance QUARTERLY; bounds 5%–30%.
# ============================================================

import pandas as pd
import numpy as np
import glob, os
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# -----------------------------
# Config
# -----------------------------
TRAIN_QUARTERS = 12
ANNUALIZE      = 4            # quarterly to annual
Z_95           = 1.645        # one-sided 95%
Z_975          = 1.96         # one-sided 97.5%
WMIN, WMAX     = 0.05, 0.30
OCEAN_BG       = "#0b1e2d"
LINE_COLOR     = "#17becf"
WATERMARK      = "Orca's Capital Strategies"

# Market caps ($B) for current mcap portfolio (order will follow R.columns)
MCAPS = {
    "NVDA": 4549.93,
    "MSFT": 3863.09,
    "AAPL": 3790.98,
    "GOOGL":2961.82,
    "AMZN": 2353.00,
    "META": 1802.06,
    "TSLA": 1527.77,
}

TICKERS = ["AAPL","AMZN","GOOGL","META","MSFT","NVDA","TSLA"]

# -----------------------------
# Loaders & utilities
# -----------------------------
def load_monthly_returns_from_change_pct(path):
    """
    Load monthly 'Change %' (e.g., 2.15%) from CSV and return a monthly return series in decimals.
    No additional transformation beyond % -> decimal.
    """
    df = pd.read_csv(path)
    if "Date" not in df.columns or "Change %" not in df.columns:
        raise ValueError(f"Need 'Date' and 'Change %' in {path}")
    df["Date"] = pd.to_datetime(df["Date"])
    r = df["Change %"].astype(str).str.replace("%","",regex=False)
    r = pd.to_numeric(r.str.replace(",","",regex=False), errors="coerce")/100.0
    s = pd.Series(r.values, index=df["Date"]).sort_index()  # monthly returns
    # Ensure monthly frequency (Date column is month-end or similar). We'll resample conservatively:
    s = s.resample("M").last()  # use last available "Change %" within each month
    return s

def monthly_to_quarterly_returns(r_monthly):
    """
    Geometric compounding from monthly to quarter-end:
    R_Q = prod(1+r_m) - 1 over months within each quarter.
    """
    return (1.0 + r_monthly).resample("QE").prod() - 1.0

def load_rf_quarterly_level(path):
    """
    Load DTB3 monthly level (percent, annual). Return QUARTERLY annual level (decimal),
    taking the LAST monthly observation within each quarter (level, not a return).
    """
    rf = pd.read_csv(path)
    dcol = [c for c in rf.columns if "date" in c.lower()][0]
    vcol = [c for c in rf.columns if c != dcol][0]
    rf[dcol] = pd.to_datetime(rf[dcol])
    rf[vcol] = pd.to_numeric(rf[vcol].replace(".", np.nan), errors="coerce")
    rf = rf.rename(columns={dcol:"Date", vcol:"DTB3"}).dropna().set_index("Date").sort_index()
    rf_m_level = rf["DTB3"] / 100.0                  # annual level in decimal
    rf_q_level = rf_m_level.resample("QE").last()    # quarter-end level
    return rf_q_level

def align_quarterly(rq_dict, rf_q_level):
    R = pd.DataFrame(rq_dict).dropna(how="any")      # strict intersection
    rfL = rf_q_level.reindex(R.index).dropna()
    common = R.index.intersection(rfL.index)
    return R.loc[common], rfL.loc[common]

def cagr(r):      return (1+r).prod()**(ANNUALIZE/len(r)) - 1 if len(r) else np.nan
def ann_vol(r):   return r.std(ddof=0)*np.sqrt(ANNUALIZE)
def max_dd(r):
    eq=(1+r.fillna(0)).cumprod(); peak=eq.cummax(); dd=eq/peak-1; return dd.min()
def var_emp(r, alpha=2.5): return np.percentile(r, alpha) if len(r) else np.nan
def win_loss(r):
    w=(r>0).sum(); l=(r<=0).sum(); return np.nan if l==0 else w/l
def calmar(r):
    mdd=max_dd(r)
    return np.nan if (np.isnan(mdd) or mdd>=0) else cagr(r)/abs(mdd)

def summarize(name, r_q, rf_level_annual_series):
    """
    Sharpe (annual) from QUARTERLY returns:
      mean_ann = mean_Q * 4
      vol_ann  = std_Q * sqrt(4)
      rf_ann   = average annual level (decimal) over evaluation span
      Sharpe   = (mean_ann - rf_ann) / vol_ann
    """
    rf_ann_mean = rf_level_annual_series.reindex(r_q.index).mean()
    mu_q  = r_q.mean()
    sig_q = r_q.std(ddof=0)
    mu_ann  = mu_q * ANNUALIZE
    sig_ann = sig_q * np.sqrt(ANNUALIZE)
    sharpe  = np.nan if sig_ann==0 else (mu_ann - rf_ann_mean)/sig_ann
    return {
        "Portfolio": name,
        "CAGR": cagr(r_q),
        "Ann Vol": sig_ann,
        "Sharpe": sharpe,
        "Max DD": max_dd(r_q),
        "VaR 97.5% (quarterly)": var_emp(r_q, 2.5),
        "Total Return": (1+r_q).prod()-1,
        "Calmar": calmar(r_q),
        "Win/Loss": win_loss(r_q),
        "Obs (quarters)": int(len(r_q))
    }

# -----------------------------
# Build QUARTERLY returns & RF LEVELS
# -----------------------------
rets_q = {}
for tkr in TICKERS:
    pats = sorted(glob.glob(f"{tkr} Historical Data*.csv"))
    if not pats: raise FileNotFoundError(f"Missing: {tkr} Historical Data*.csv")
    r_m = load_monthly_returns_from_change_pct(pats[0])
    rets_q[tkr] = monthly_to_quarterly_returns(r_m)

rf_level_q_annual = load_rf_quarterly_level("DTB3.csv")  # decimal annual level per quarter
R, rfL = align_quarterly(rets_q, rf_level_q_annual)
assets = list(R.columns); n=len(assets)

# Fixed portfolios
mc = np.array([MCAPS[a] for a in assets], dtype=float)
w_mcap_fixed = mc / mc.sum()
w_equal_fixed= np.ones(n)/n

# -----------------------------
# Optimization helpers (Quarterly)
# -----------------------------
bounds = [(WMIN, WMAX)]*n

def slsqp_sum1_bounds_maximize(objfun, bounds):
    def minfun(w): return -objfun(w)
    cons = ({'type':'eq','fun':lambda w: w.sum()-1.0},)
    w0 = np.ones(n)/n
    res = minimize(minfun, w0, method="SLSQP", bounds=tuple(bounds), constraints=cons,
                   options={'ftol':1e-12,'maxiter':500})
    if res.success: return res.x
    lo = np.array([b[0] for b in bounds]); hi = np.array([b[1] for b in bounds])
    w = np.clip(w0, lo, hi); return w / w.sum()

def optimize_sharpe(mu_ex_q_vec, cov_q, bounds):
    """Maximize portfolio Sharpe at the QUARTERLY step: (mu_ex' w) / sqrt(w' Σ w)."""
    def obj(w):
        num = mu_ex_q_vec @ w
        den = np.sqrt(max(w @ cov_q @ w, 1e-12))
        return -(num/den)
    cons = ({'type':'eq','fun':lambda w: w.sum()-1.0},)
    w0 = np.ones(n)/n
    res = minimize(obj, w0, method="SLSQP", bounds=tuple(bounds), constraints=cons,
                   options={'ftol':1e-12,'maxiter':500})
    if res.success: return res.x
    lo = np.array([b[0] for b in bounds]); hi = np.array([b[1] for b in bounds])
    w = np.clip(w0, lo, hi); return w/w.sum()

def mvar_vector(cov_df, w_prev, z):
    """Δ-normal MVaR per asset at level z using previous weights and window Σ."""
    cov = cov_df.values
    sigma_p = np.sqrt(max(w_prev @ cov @ w_prev, 1e-12))
    contrib = cov @ w_prev
    mvar = z * (contrib / max(sigma_p,1e-12))
    return pd.Series(mvar, index=cov_df.columns)

def var_ordinal_vec(window_R, q):  # q = 0.05 (95%) or 0.025 (97.5%)
    """Standalone empirical VaR per asset at quantile q (positive loss)."""
    return -window_R.quantile(q, axis=0)

def optimize_score_linear(scores, bounds):
    s = np.asarray(scores, float)
    def objfun(w): return np.dot(s, w)
    return slsqp_sum1_bounds_maximize(objfun, bounds)

# -----------------------------
# Backtests (Quarterly, all rules)
# -----------------------------
def run_fixed_quarterly(target_w):
    start = TRAIN_QUARTERS
    pr = []
    for t in range(start, len(R)-1):
        pr.append(float(target_w @ R.iloc[t+1].values))
    idx = R.index[start+1:]
    W = pd.DataFrame([target_w]*len(idx), index=idx, columns=assets)
    return pd.Series(pr, index=idx), W

def run_sharpe_quarterly():
    """
    Quarterly regression on QUARTERLY returns.
    Excess QUARTERLY mean = mean(Rq) - rf_level/4 (rf is a level).
    """
    start=TRAIN_QUARTERS
    pr=[]; W=[]
    for t in range(start, len(R)-1):
        win=slice(t-TRAIN_QUARTERS, t)
        Rw = R.iloc[win]
        rf_q_level_mean = rfL.iloc[win].mean()            # annual level (decimal)
        rf_per_q = rf_q_level_mean / ANNUALIZE            # per quarter level for excess
        mu_ex_q_vec = (Rw.mean().values - rf_per_q)       # vector per asset
        cov_q = Rw.cov().values
        w = optimize_sharpe(mu_ex_q_vec, cov_q, bounds)
        pr.append(float(w @ R.iloc[t+1].values))
        W.append(pd.Series(w, index=assets))
    idx=R.index[start+1:]
    return pd.Series(pr, index=idx), pd.DataFrame(W, index=idx)

def run_er_mvar_score_quarterly(z):
    """
    Score_i = mu_ex_i / MVaR_i (quarterly mu_ex, quarterly MVaR using w_prev).
    """
    start=TRAIN_QUARTERS
    pr=[]; W=[]; w_prev = np.ones(n)/n
    for t in range(start, len(R)-1):
        win=slice(t-TRAIN_QUARTERS, t)
        Rw = R.iloc[win]
        rf_q_level_mean = rfL.iloc[win].mean()
        rf_per_q = rf_q_level_mean / ANNUALIZE
        mu_ex = (Rw.mean() - rf_per_q)                    # pd.Series
        mvar  = mvar_vector(Rw.cov(), w_prev, z)          # pd.Series
        mvar = mvar.replace([np.inf,-np.inf], np.nan).abs()
        mvar = mvar.where(mvar>0, np.nan)
        scores = (mu_ex / mvar).replace([np.inf,-np.inf], np.nan).fillna(0.0).values
        w = optimize_score_linear(scores, bounds)
        pr.append(float(w @ R.iloc[t+1].values))
        W.append(pd.Series(w, index=assets))
        w_prev = w
    idx=R.index[start+1:]
    return pd.Series(pr, index=idx), pd.DataFrame(W, index=idx)

def run_er_var_score_quarterly(q):
    """
    Score_i = mu_ex_i / VaR_i (standalone empirical VaR at quantile q, quarterly data).
    """
    start=TRAIN_QUARTERS
    pr=[]; W=[]
    for t in range(start, len(R)-1):
        win=slice(t-TRAIN_QUARTERS, t)
        Rw = R.iloc[win]
        rf_q_level_mean = rfL.iloc[win].mean()
        rf_per_q = rf_q_level_mean / ANNUALIZE
        mu_ex = (Rw.mean() - rf_per_q)                    # pd.Series
        var_i = var_ordinal_vec(Rw, q)                    # pd.Series (positive)
        var_i = var_i.replace([np.inf,-np.inf], np.nan)
        var_i = var_i.where(var_i>0, np.nan)
        scores = (mu_ex / var_i).replace([np.inf,-np.inf], np.nan).fillna(0.0).values
        w = optimize_score_linear(scores, bounds)
        pr.append(float(w @ R.iloc[t+1].values))
        W.append(pd.Series(w, index=assets))
    idx=R.index[start+1:]
    return pd.Series(pr, index=idx), pd.DataFrame(W, index=idx)

# -----------------------------
# Run all portfolios
# -----------------------------
port_eq,      W_eq      = run_fixed_quarterly(w_equal_fixed)
port_mcap,    W_mcap    = run_fixed_quarterly(w_mcap_fixed)
port_shp,     W_shp     = run_sharpe_quarterly()

# ER/MVaR scores (MVaR uses w_prev) at 95% and 97.5%
port_mvar95,  W_mvar95  = run_er_mvar_score_quarterly(Z_95)
port_mvar975, W_mvar975 = run_er_mvar_score_quarterly(Z_975)

# ER/Standalone VaR scores at 95% and 97.5%
port_var95,   W_var95   = run_er_var_score_quarterly(0.05)
port_var975,  W_var975  = run_er_var_score_quarterly(0.025)

# -----------------------------
# Summaries (Sharpe uses RF LEVEL annual)
# -----------------------------
def fmt_pct(x):   return "—" if pd.isna(x) else f"{x*100:.2f}%"
def fmt_ratio(x): return "—" if pd.isna(x) else f"{x:.2f}"
def fmt_var(x):   return "—" if pd.isna(x) else f"{x*100:.2f}%"

summ = []
def add_sum(name, r):
    summ.append(summarize(name, r, rfL))

add_sum("Equal-Weight (Quarterly)", port_eq)
add_sum("Current Market-Cap (Quarterly)", port_mcap)
add_sum("Sharpe-Optimized (5–30%, Quarterly)", port_shp)

add_sum("ER/MVaR Score — 95% (5–30%, Quarterly)",  port_mvar95)
add_sum("ER/MVaR Score — 97.5% (5–30%, Quarterly)",port_mvar975)

add_sum("ER/Standalone VaR Score — 95% (5–30%, Quarterly)",  port_var95)
add_sum("ER/Standalone VaR Score — 97.5% (5–30%, Quarterly)",port_var975)

summary_df = pd.DataFrame(summ)
print("\n=== Portfolio Summary (OOS after 12-quarter training) ===")
out = summary_df.copy()
out["CAGR"]                 = out["CAGR"].apply(fmt_pct)
out["Ann Vol"]              = out["Ann Vol"].apply(fmt_pct)
out["Sharpe"]               = out["Sharpe"].round(2)
out["Max DD"]               = out["Max DD"].apply(fmt_pct)
out["VaR 97.5% (quarterly)"]= out["VaR 97.5% (quarterly)"].apply(fmt_var)
out["Total Return"]         = out["Total Return"].apply(fmt_pct)
out["Calmar"]               = out["Calmar"].apply(fmt_ratio)
out["Win/Loss"]             = out["Win/Loss"].apply(fmt_ratio)
print(out[["Portfolio","CAGR","Ann Vol","Sharpe","Max DD","VaR 97.5% (quarterly)",
           "Total Return","Calmar","Win/Loss","Obs (quarters)"]].to_string(index=False))

# -----------------------------
# Optional: quick last-weights view
# -----------------------------
def show_last(name, W):
    print(f"\n=== Last Weights — {name} ===")
    print(W.iloc[-1].round(4).to_string())

show_last("Equal-Weight", W_eq)
show_last("Current Market-Cap", W_mcap)
show_last("Sharpe-Optimized", W_shp)
show_last("ER/MVaR Score — 95%",  W_mvar95)
show_last("ER/MVaR Score — 97.5%",W_mvar975)
show_last("ER/Standalone VaR Score — 95%",  W_var95)
show_last("ER/Standalone VaR Score — 97.5%",W_var975)

# -----------------------------
# Oceanic visuals (optional)
# -----------------------------
def ocean_axes(ax, title):
    ax.set_facecolor(OCEAN_BG); ax.figure.set_facecolor(OCEAN_BG)
    ax.set_title(title, color="white", fontsize=15, fontweight="bold", pad=18)
    ax.tick_params(colors="white")
    ax.spines["bottom"].set_color("white"); ax.spines["left"].set_color("white")
    ax.spines["top"].set_visible(False);   ax.spines["right"].set_visible(False)
    ax.grid(alpha=0.2, color="white")

def add_watermark(ax):
    ax.text(0.5, 0.05, WATERMARK, transform=ax.transAxes,
            fontsize=16, color="white", alpha=0.15,
            ha="center", va="center", rotation=30)

def equity_curve(r, title):
    eq=(1+r.fillna(0)).cumprod()
    fig, ax = plt.subplots(figsize=(12,6))
    ax.plot(eq.index, eq.values, color=LINE_COLOR, linewidth=1.8, label=title)
    ocean_axes(ax, f"Equity Curve — {title}"); add_watermark(ax)
    ax.set_ylabel("Growth of $1", color="white")
    last_date, last_val = eq.index[-1], eq.iloc[-1]
    ax.scatter([last_date],[last_val], color="red")
    ax.annotate(f"{last_val:.2f}x", xy=(last_date,last_val), xytext=(8,8),
                textcoords="offset points", fontsize=10, color="white", weight="bold")
    ax.legend(facecolor=OCEAN_BG, edgecolor="none", labelcolor="white")
    plt.tight_layout(); plt.show()

def stacked_weights(W, title):
    fig, ax = plt.subplots(figsize=(12,6))
    W.plot(kind="area", ax=ax, linewidth=0)
    ocean_axes(ax, title); add_watermark(ax)
    ax.set_ylabel("Weight", color="white"); ax.set_ylim(0,1)
    plt.tight_layout(); plt.show()

# Example plots (comment out when running headless)
equity_curve(port_eq,      "Equal-Weight (Quarterly)")
equity_curve(port_mcap,    "Current Market-Cap (Quarterly)")
equity_curve(port_shp,     "Sharpe-Optimized (5–30%, Quarterly)")
equity_curve(port_mvar95,  "ER/MVaR Score — 95% (5–30%, Quarterly)")
equity_curve(port_mvar975, "ER/MVaR Score — 97.5% (5–30%, Quarterly)")
equity_curve(port_var95,   "ER/Standalone VaR Score — 95% (5–30%, Quarterly)")
equity_curve(port_var975,  "ER/Standalone VaR Score — 97.5% (5–30%, Quarterly)")

stacked_weights(W_shp,     "Weights Over Time — Sharpe (Quarterly)")
stacked_weights(W_mvar95,  "Weights Over Time — ER/MVaR Score 95% (Quarterly)")
stacked_weights(W_mvar975, "Weights Over Time — ER/MVaR Score 97.5% (Quarterly)")
stacked_weights(W_var95,   "Weights Over Time — ER/Standalone VaR Score 95% (Quarterly)")
stacked_weights(W_var975,  "Weights Over Time — ER/Standalone VaR Score 97.5% (Quarterly)")



















```
