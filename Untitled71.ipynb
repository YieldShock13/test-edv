```python
# ==========================================================
# Dual-Signal Strategy (Monthly 1st)
# Enter only when BOTH z-scores agree:
#   futures z < 0 & yield z < 0  -> Long DE futures
#   futures z > 0 & yield z > 0  -> Long US futures
# Else flat.
#
# Also rebuilds:
#    Yield spread (bps) + z
#    Futures spreads: 11 and OLS-hedged + z
# Charts:
#    Annotated last points on spreads
#    Entry markers on futures price chart
#    Equity & Drawdown
#    Monthly returns bars (wins/losses)
# Prints per-trade P&L table.
# Watermark: "Orca's Capital Strategies"
# ==========================================================

import os, glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import statsmodels.api as sm

# ---------------- Config ----------------
WATERMARK = "Orca's Capital Strategies"
DATE_FMT_M = DateFormatter("%Y-%m")
FUT_Z_SOURCE = "1x1"   # "1x1" uses z of (DEUS); set "hedged" to use z of (DEUS)
ROLLING_Z_WINDOW = None

# Exact filenames
FN_DE_YLD = "Germany 2-Year Bond Yield Historical Data (2).csv"
FN_US_YLD = "United States 2-Year Bond Yield Historical Data (1).csv"
FN_DE_FUT = "Euro SCHATZ Futures Historical Data.csv"
FN_US_FUT = "US 2 Year T-Note Futures Historical Data.csv"
FN_RF_3M  = "United States 3-Month Bond Yield Historical Data.csv"  # optional; not needed for TRR

# ---------------- Helpers ----------------
def resolve_path(filename):
    for p in (f"/content/{filename}", f"/mnt/data/{filename}", filename):
        if os.path.isfile(p): return p
    for base in ("/content", "/mnt/data", "."):
        hits = glob.glob(os.path.join(base, "**", filename), recursive=True)
        if hits: return hits[0]
    raise FileNotFoundError(f"Could not find: {filename}")

def robust_date(s: pd.Series) -> pd.Series:
    s = s.astype(str)
    d1 = pd.to_datetime(s, dayfirst=True,  errors="coerce")
    d2 = pd.to_datetime(s, dayfirst=False, errors="coerce")
    return d1 if d1.max() >= d2.max() else d2

def parse_number(series: pd.Series) -> pd.Series:
    s = series.astype(str).str.strip().str.replace("%","",regex=False)
    smpl = s.sample(min(200, len(s)), random_state=0)
    comma_no_dot = ((smpl.str.contains(",")) & ~smpl.str.contains(r"\.")).mean()
    if comma_no_dot > 0.6:
        s = s.str.replace(".", "", regex=False)   # thousands dot
        s = s.str.replace(",", ".", regex=False)  # decimal comma -> dot
    else:
        s = s.str.replace(",", "", regex=False)   # thousands comma
    return pd.to_numeric(s, errors="coerce")

def load_monthly_first(path: str, colname: str) -> pd.Series:
    df = pd.read_csv(path)
    df.columns = [c.strip() for c in df.columns]
    if "Date" not in df.columns or "Price" not in df.columns:
        raise ValueError(f"'Date' and 'Price' required in {path}. Found: {df.columns.tolist()}")
    df["Date"]  = robust_date(df["Date"])
    df["Price"] = parse_number(df["Price"])
    df = df.dropna(subset=["Date","Price"]).sort_values("Date").set_index("Date")
    # Enforce one obs per month (your files are stamped on the 1st)
    df = df[~df.index.duplicated(keep="first")]
    return df["Price"].rename(colname)

def add_wm(fig): fig.text(0.01, 0.01, WATERMARK, fontsize=12, alpha=0.35, ha="left", va="bottom")

def annotate_last(ax, x, y, fmt="{:.2f}", dx=8, dy=0):
    ax.annotate(fmt.format(y[-1]), xy=(x[-1], y[-1]),
                xytext=(dx, dy), textcoords="offset points", va="center")

# ---------------- Load data ----------------
p_de_yld = resolve_path(FN_DE_YLD)
p_us_yld = resolve_path(FN_US_YLD)
p_de_fut = resolve_path(FN_DE_FUT)
p_us_fut = resolve_path(FN_US_FUT)

DE2Y = load_monthly_first(p_de_yld, "DE2Y_yld")
US2Y = load_monthly_first(p_us_yld, "US2Y_yld")
DE_F = load_monthly_first(p_de_fut, "DE_fut")
US_F = load_monthly_first(p_us_fut, "US_fut")

# Align indices
ix_yld = DE2Y.index.intersection(US2Y.index)
yld = pd.concat([DE2Y.reindex(ix_yld), US2Y.reindex(ix_yld)], axis=1).dropna()

ix_fut = DE_F.index.intersection(US_F.index)
fut = pd.concat([DE_F.reindex(ix_fut), US_F.reindex(ix_fut)], axis=1).dropna()

# ---------------- Yield spread (bps) + z ----------------
yld["Spread_bps"] = (yld["DE2Y_yld"] - yld["US2Y_yld"]) * 100.0
mu_y, sd_y = yld["Spread_bps"].mean(), yld["Spread_bps"].std(ddof=0)
yld["Z_yield"] = (yld["Spread_bps"] - mu_y) / sd_y if sd_y != 0 else np.nan

# ---------------- Futures spreads + z ----------------
# 11 spread
fut["Spread_1x1"] = fut["DE_fut"] - fut["US_fut"]
mu_1x1, sd_1x1 = fut["Spread_1x1"].mean(), fut["Spread_1x1"].std(ddof=0)
fut["Z_1x1"] = (fut["Spread_1x1"] - mu_1x1) / sd_1x1 if sd_1x1 != 0 else np.nan

# OLS-hedged
X = sm.add_constant(fut["US_fut"])
y = fut["DE_fut"]
ols = sm.OLS(y, X).fit()
beta = float(ols.params["US_fut"]); alpha = float(ols.params["const"])
fut["Spread_hedged"] = fut["DE_fut"] - beta * fut["US_fut"]
mu_h, sd_h = fut["Spread_hedged"].mean(), fut["Spread_hedged"].std(ddof=0)
fut["Z_hedged"] = (fut["Spread_hedged"] - mu_h) / sd_h if sd_h != 0 else np.nan

# ---------------- Strategy (dual-signal, rebalance monthly on 1st) ----------------
# Forward returns (signal at t, hold to t+1)
DE_ret_fwd = DE_F.shift(-1) / DE_F - 1.0
US_ret_fwd = US_F.shift(-1) / US_F - 1.0
ret_df = pd.concat([DE_ret_fwd.rename("DE_ret_fwd"), US_ret_fwd.rename("US_ret_fwd")], axis=1)

# Merge futures z's, yield z, and next-month returns on common months
panel = fut.join(yld[["Z_yield"]], how="inner").join(ret_df, how="inner").dropna()

fut_z_col = "Z_1x1" if FUT_Z_SOURCE.lower() == "1x1" else "Z_hedged"
panel["FutZ"] = panel[fut_z_col]
panel["YldZ"] = panel["Z_yield"]

# Dual-signal rule
long_DE = (panel["FutZ"] < 0) & (panel["YldZ"] < 0)
long_US = (panel["FutZ"] > 0) & (panel["YldZ"] > 0)
flat    = ~(long_DE | long_US)

# One-month return depending on entry
panel["Strat_ret"] = 0.0
panel.loc[long_DE, "Strat_ret"] = panel.loc[long_DE, "DE_ret_fwd"]
panel.loc[long_US, "Strat_ret"] = panel.loc[long_US, "US_ret_fwd"]

# Create per-trade log (only months with a position)
trades = panel.loc[long_DE | long_US, ["FutZ","YldZ","DE_ret_fwd","US_ret_fwd","Strat_ret"]].copy()
trades["Leg"] = np.where(long_DE.loc[trades.index], "DE", "US")
trades.rename(columns={"DE_ret_fwd":"DE_ret", "US_ret_fwd":"US_ret"}, inplace=True)

# Performance series
inc = panel["Strat_ret"]
equity = (1 + inc).cumprod()
roll_max = equity.cummax()
drawdown = equity / roll_max - 1.0

# Metrics
cum_gain = equity.iloc[-1] - 1.0
wins = (inc > 0).sum(); losses = (inc < 0).sum()
win_loss_ratio = (wins / losses) if losses > 0 else np.nan
var95 = -np.nanpercentile(inc.dropna(), 5)       # 1-month VaR 95%
ann_vol = inc.std(ddof=0) * np.sqrt(12)          # Annualized volatility
n = len(inc)
cagr = equity.iloc[-1]**(12 / n) - 1 if n > 0 else np.nan
total_risk_ratio = (cagr / ann_vol) if (ann_vol and ann_vol != 0) else np.nan
max_dd = drawdown.min()

# ---------------- Charts (annotate last points & mark entries) ----------------
def finalize(fig): add_wm(fig); plt.tight_layout(); plt.show()

# 1) Yield spread (bps) with last label
fig1, ax1 = plt.subplots(figsize=(11,5.0))
ax1.plot(yld.index, yld["Spread_bps"])
ax1.set_title("Yield Spread (DE2Y  US2Y)  Monthly (1st)")
ax1.set_ylabel("bps"); ax1.set_xlabel("Date"); ax1.grid(True, alpha=0.3)
ax1.xaxis.set_major_formatter(DATE_FMT_M)
annotate_last(ax1, yld.index, yld["Spread_bps"].values, fmt="{:.1f} bps")
finalize(fig1)

# 2) Futures prices with ENTRY MARKERS
fig2, ax2 = plt.subplots(figsize=(12,5.6))
ax2.plot(fut.index, fut["DE_fut"], label="DE futures")
ax2.plot(fut.index, fut["US_fut"], label="US futures")
# Entry markers
de_entries = panel.index[long_DE]
us_entries = panel.index[long_US]
ax2.scatter(de_entries, fut.loc[de_entries, "DE_fut"], marker="^", s=90, label="ENTRY: Long DE", zorder=5)
ax2.scatter(us_entries, fut.loc[us_entries, "US_fut"], marker="v", s=90, label="ENTRY: Long US", zorder=5)
ax2.set_title(f"Futures (Monthly 1st)  Entries when BOTH z's agree  |  FutZ source: {FUT_Z_SOURCE}")
ax2.set_ylabel("Price"); ax2.set_xlabel("Date"); ax2.grid(True, alpha=0.3)
ax2.xaxis.set_major_formatter(DATE_FMT_M)
annotate_last(ax2, fut.index, fut["DE_fut"].values, fmt="{:.2f}")
annotate_last(ax2, fut.index, fut["US_fut"].values, fmt="{:.2f}")
ax2.legend()
finalize(fig2)

# 3) Futures 11 spread (last label)
fig3, ax3 = plt.subplots(figsize=(11,5.0))
ax3.plot(fut.index, fut["Spread_1x1"], label="Spread 11 (DE  US)")
ax3.axhline(mu_1x1, linestyle="--", alpha=0.8)
ax3.axhline(mu_1x1 + 2*sd_1x1, linestyle=":", alpha=0.8)
ax3.axhline(mu_1x1 - 2*sd_1x1, linestyle=":", alpha=0.8)
ax3.set_title("Spread 11 with 2")
ax3.set_ylabel("Price units"); ax3.set_xlabel("Date"); ax3.grid(True, alpha=0.3)
ax3.xaxis.set_major_formatter(DATE_FMT_M)
annotate_last(ax3, fut.index, fut["Spread_1x1"].values, fmt="{:.3f}")
ax3.legend()
finalize(fig3)

# 4) Equity curve + last label
fig4, ax4 = plt.subplots(figsize=(11,5.0))
ax4.plot(panel.index, equity, label="Equity")
ax4.set_title("Equity Curve  Dual-Signal Strategy (Monthly, 1st)")
ax4.set_ylabel("Growth of $1"); ax4.set_xlabel("Date"); ax4.grid(True, alpha=0.3)
ax4.xaxis.set_major_formatter(DATE_FMT_M)
annotate_last(ax4, panel.index, equity.values, fmt="{:.3f}")
ax4.legend()
finalize(fig4)

# 5) Monthly returns bars (wins vs losses)
fig5, ax5 = plt.subplots(figsize=(11,4.2))
colors = np.where(inc >= 0, "tab:blue", "tab:orange")
ax5.bar(panel.index, inc.values, width=20, align="center", color=colors)
ax5.axhline(0, linewidth=1)
ax5.set_title("Monthly Strategy Returns  Wins (blue) vs Losses (orange)")
ax5.set_ylabel("Return"); ax5.set_xlabel("Date"); ax5.grid(True, alpha=0.25)
ax5.xaxis.set_major_formatter(DATE_FMT_M)
finalize(fig5)

# 6) Drawdown
fig6, ax6 = plt.subplots(figsize=(11,3.8))
ax6.plot(panel.index, drawdown)
ax6.set_title("Drawdown")
ax6.set_ylabel("Drawdown"); ax6.set_xlabel("Date"); ax6.grid(True, alpha=0.3)
ax6.xaxis.set_major_formatter(DATE_FMT_M)
finalize(fig6)

# ---------------- Per-trade P&L table ----------------
trades_out = trades[["Leg","FutZ","YldZ","DE_ret","US_ret","Strat_ret"]].copy()
trades_out.index.name = "Date"
print("\n=== Per-Trade P&L (only months with a position) ===")
print(trades_out.to_string(float_format=lambda x: f"{x:,.4f}"))

# ---------------- Metrics ----------------
print("\n=== Dual-Signal Strategy Metrics ===")
print(f"Futures z-source: {FUT_Z_SOURCE}")
print(f"Months: {len(panel)}  |  Active months: {len(trades_out)}")
print(f"Cumulative Gain: {cum_gain:.2%}")
print(f"Max Drawdown: {max_dd:.2%}")
print(f"Win/Loss ratio: {win_loss_ratio:.3f}  (Wins={wins}, Losses={losses})")
print(f"VaR 95% (1-month): {var95:.2%}")
print(f"Annualized Volatility: {ann_vol:.2%}")
print(f"CAGR: {cagr:.2%}")
print(f"Total Risk Ratio (CAGR / Ann.Vol): {total_risk_ratio:.3f}")

# ---------------- Save outputs ----------------
out_all = pd.concat(
    [
        yld[["Spread_bps","Z_yield"]],
        fut[["Spread_1x1","Z_1x1","Spread_hedged","Z_hedged"]],
        panel[["FutZ","YldZ","DE_ret_fwd","US_ret_fwd","Strat_ret"]],
        equity.rename("Equity"),
        drawdown.rename("Drawdown")
    ],
    axis=1
)
out_all.to_csv("/content/dual_signal_strategy_output.csv")
print("\nSaved CSV -> /content/dual_signal_strategy_output.csv")





```
