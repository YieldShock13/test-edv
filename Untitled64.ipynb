```python
# ======================================================
# TRANSMISSION AFTER +2σ OAS — ALIGNMENT-AWARE (FINAL)
# Auto-picks week anchor to maximize HYG~HY_OAS event R² (lag0)
# Units: assets = %/week (log rtns*100); OAS = Δbps
# Outputs:
#  - Auto anchor choice + sanity dashboard (event R² by asset)
#  - Event calendars (shock dates & sizes)
#  - Weekly β(h) tables (β, se_HAC, t, p, R2_full, AdjR2_full, R2_event, N) + plots
#  - Multi-lag OLS (0..12): ANOVA with partial R², model R²/AdjR²/N; HAC coef/t/p + plot
# ======================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.stats.anova import anova_lm

# ---------- FILES (update paths if needed) ----------
FILES = {
    "SPY":  "/content/SPY Historical Data (3).csv",
    "HYG":  "/content/HYG Historical Data (1).csv",
    "Gold": "/content/Gold Futures Historical Data.csv",
    "WTI":  "/content/Crude Oil WTI Futures Historical Data.csv",
    "HY_OAS": "/content/BAMLH0A0HYM2 (9).csv",
    "IG_OAS": "/content/BAMLC0A0CM (2).csv",
}

ASSET_LIST = ["SPY","HYG","Gold","WTI"]

# ---------- LOADERS ----------
def load_generic_csv(path, rename_as):
    df = pd.read_csv(path)
    dcols = [c for c in df.columns if "date" in c.lower()]
    if not dcols: raise ValueError(f"No date column in {path}")
    dcol = dcols[0]
    df[dcol] = pd.to_datetime(df[dcol])
    df = df.sort_values(dcol).set_index(dcol)
    vcol = None
    for c in df.columns:
        ser = pd.to_numeric(df[c].astype(str).str.replace(",",""), errors="coerce")
        if ser.notna().sum() > 0:
            df[c] = ser
            vcol = c
            break
    if vcol is None: raise ValueError(f"No numeric column in {path}")
    return df[[vcol]].rename(columns={vcol: rename_as})

def pct_ret(x): return np.log(x).diff() * 100.0  # % per week

def to_bps(s):
    med = s.median()
    if pd.notna(med) and 0 < med < 20:  # looks like %
        return s * 100.0
    return s

def prep_with_anchor(anchor: str):
    """
    1) Load daily
    2) For OAS: upsample to BusinessDay and ffill (to avoid Friday gaps), then resample weekly.
       For assets: resample weekly.
    3) Compute weekly % returns and Δbps.
    """
    # Assets daily -> weekly
    assets_daily = {k: load_generic_csv(FILES[k], k) for k in ASSET_LIST}
    assets_w = {k: assets_daily[k].resample(anchor).last() for k in ASSET_LIST}
    prices_w = pd.concat([assets_w[k] for k in ASSET_LIST], axis=1)
    rets_w   = prices_w.apply(pct_ret)

    # OAS: daily -> BusinessDay -> ffill -> weekly
    hy_d = load_generic_csv(FILES["HY_OAS"], "HY_OAS_raw").asfreq("B").ffill()
    ig_d = load_generic_csv(FILES["IG_OAS"], "IG_OAS_raw").asfreq("B").ffill()
    hy_w = to_bps(hy_d["HY_OAS_raw"]).rename("HY_OAS").resample(anchor).last()
    ig_w = to_bps(ig_d["IG_OAS_raw"]).rename("IG_OAS").resample(anchor).last()
    oas_w = pd.concat([hy_w, ig_w], axis=1)

    # Weekly Δbps
    doas_w = oas_w.diff()

    # Merge
    data = pd.concat([rets_w, doas_w], axis=1).dropna().copy()
    return data

def build_shocks(series, sigma=2.0):
    thr = series.mean() + sigma*series.std()
    flag = (series > thr).astype(int)
    size = series.where(flag==1, 0.0)
    return flag.rename(series.name+"_Shock"), size.rename(series.name+"_ShockSize"), thr

def event_r2(y, S):
    df = pd.DataFrame({"y": y, "S": S}).dropna()
    df_ev = df[df["S"]!=0]
    if len(df_ev) < 8: return np.nan
    X = sm.add_constant(df_ev["S"])
    ols = sm.OLS(df_ev["y"], X).fit()
    return float(ols.rsquared)

# ---------- AUTO-SELECT ANCHOR ----------
anchors = ["W-MON","W-TUE","W-WED","W-THU","W-FRI"]
anchor_scores = []
for a in anchors:
    D = prep_with_anchor(a)
    HY_Shock, HY_S, _ = build_shocks(D["HY_OAS"])
    r2 = event_r2(D["HYG"], HY_S)
    anchor_scores.append((a, r2, len(D)))
best_anchor = max(anchor_scores, key=lambda x: (np.nan_to_num(x[1], nan=-1), x[2]))[0]
print("Auto-selected week anchor:", best_anchor, "| scores:", anchor_scores)

# ---------- PREP DATA WITH BEST ANCHOR ----------
data = prep_with_anchor(best_anchor)
HY_Shock, HY_S, hy_thr = build_shocks(data["HY_OAS"])
IG_Shock, IG_S, ig_thr = build_shocks(data["IG_OAS"])
data = pd.concat([data, HY_Shock, HY_S, IG_Shock, IG_S], axis=1)

# ---------- SANITY DASHBOARD ----------
def dashboard():
    rows = []
    for lab, Sname in [("HY","HY_OAS_ShockSize"),("IG","IG_OAS_ShockSize")]:
        for asset in ASSET_LIST:
            r2 = event_r2(data[asset], data[Sname])
            rows.append({"Shock": lab, "Asset": asset, "Event_R2_lag0": round(r2,4) if pd.notna(r2) else np.nan})
    print("\n=== Event-only R² (lag0) dashboard ===")
    print(pd.DataFrame(rows).pivot(index="Asset", columns="Shock", values="Event_R2_lag0").to_string())
dashboard()

# ---------- EVENT CALENDARS ----------
def print_event_calendar(label, shock_flag, delta_col, topn=25):
    ev = pd.DataFrame({"ΔOAS_bps": data[delta_col], "Shock": shock_flag}).query("Shock==1")[["ΔOAS_bps"]]
    print(f"\n=== {label} +2σ ΔOAS Shock Dates (top {topn}) — anchor {best_anchor} ===")
    if ev.empty:
        print("(No shocks found)")
    else:
        print(ev.sort_values("ΔOAS_bps", ascending=False).head(topn).to_string())

print_event_calendar("HY", HY_Shock, "HY_OAS")
print_event_calendar("IG", IG_Shock, "IG_OAS")

# ======================================================
# A) WEEK-BY-WEEK TRANSMISSION β(h): y_{t+h} = α + β(h)*S_t + ε
# ======================================================
def weekly_path(asset_col, S, H=12):
    rows=[]
    for h in range(H+1):
        df = pd.DataFrame({"y": data[asset_col].shift(-h), "S": S}).dropna()
        X = sm.add_constant(df["S"])
        hac = sm.OLS(df["y"], X).fit(cov_type="HAC", cov_kwds={"maxlags":4})
        ols = sm.OLS(df["y"], X).fit()
        # Event-only R² at this horizon (using same df but keep only S!=0 rows)
        df_ev = df[df["S"]!=0]
        if len(df_ev)>=8:
            ols_ev = sm.OLS(df_ev["y"], sm.add_constant(df_ev["S"])).fit()
            r2_event = float(ols_ev.rsquared)
        else:
            r2_event = np.nan
        rows.append({
            "h":h,
            "beta_%perbp": float(hac.params["S"]),
            "se_hac": float(hac.bse["S"]),
            "t_hac": float(hac.tvalues["S"]),
            "p_hac": float(hac.pvalues["S"]),
            "R2_full": float(ols.rsquared),
            "AdjR2_full": float(ols.rsquared_adj),
            "R2_event": r2_event,
            "N": int(hac.nobs),
        })
    out = pd.DataFrame(rows).set_index("h")
    sig = out.index[out["p_hac"]<0.05]
    horizon = int(sig.max()) if len(sig) else 0
    return out, horizon

def plot_weekly_path(df_path, title):
    lo = df_path["beta_%perbp"] - 1.96*df_path["se_hac"]
    hi = df_path["beta_%perbp"] + 1.96*df_path["se_hac"]
    plt.figure(figsize=(9,5))
    df_path["beta_%perbp"].plot(marker="o", title=title)
    plt.fill_between(df_path.index.values, lo.values, hi.values, alpha=0.2)
    plt.axhline(0, ls="--", c="k"); plt.xlabel("Weeks after shock (h)"); plt.ylabel("β(h): % per 1 bp")
    plt.show()

summary_rows=[]
for lab, Sname in [("HY","HY_OAS_ShockSize"),("IG","IG_OAS_ShockSize")]:
    S = data[Sname]
    for asset in ASSET_LIST:
        path, horizon = weekly_path(asset, S, H=12)
        print(f"\n=== Weekly Transmission Path β(h): {asset} on {lab} ΔOAS shocks (anchor {best_anchor}) ===")
        print(path.to_string())
        print(f"Transmission horizon (p<0.05): {horizon} week(s)")
        plot_weekly_path(path, f"{asset}: β(h) after {lab} +2σ ΔOAS shock (0–12w)")

        peak_h = int(path["t_hac"].abs().idxmax())
        summary_rows.append({
            "ShockSource": lab, "Asset": asset,
            "Peak_h": peak_h,
            "Peak_beta_%perbp": round(path.loc[peak_h,"beta_%perbp"],4),
            "Peak_t": round(path.loc[peak_h,"t_hac"],3),
            "Peak_p": round(path.loc[peak_h,"p_hac"],5),
            "Horizon_weeks": horizon,
            "R2_event_at_peak": round(path.loc[peak_h,"R2_event"],4) if pd.notna(path.loc[peak_h,"R2_event"]) else np.nan,
            "R2_full_at_peak": round(path.loc[peak_h,"R2_full"],4),
            "AdjR2_full_at_peak": round(path.loc[peak_h,"AdjR2_full"],4),
            "N_at_peak": int(path.loc[peak_h,"N"])
        })

summary = pd.DataFrame(summary_rows)
print("\n=== Transmission Summary (week-by-week; units % per 1 bp; anchor {}) ===".format(best_anchor))
print(summary.to_string(index=False))

# ======================================================
# B) MULTI-LAG OLS (0..12): y_t = α + Σ γ_k S_{t-k} + ε
# ======================================================
def fit_multilag(asset_col, S, label, L=12):
    Xlags = pd.DataFrame({f"S_lag{k}": S.shift(k) for k in range(L+1)})
    df = pd.concat([data[asset_col].rename("y"), Xlags], axis=1).dropna()

    # Classic OLS for ANOVA / model R²
    rhs = " + ".join([f"S_lag{k}" for k in range(L+1)])
    ols_model = smf.ols(f"y ~ {rhs}", data=df).fit()
    anova = anova_lm(ols_model, typ=2).copy()
    # partial R² per lag
    SSE = float(ols_model.ssr)
    anova["partial_R2"] = anova["sum_sq"] / (anova["sum_sq"] + SSE)

    print(f"\n=== ANOVA: {asset_col} on shocked {label} ΔOAS (lags 0..{L}; anchor {best_anchor}) ===")
    print(anova)
    print(f"Model R²: {ols_model.rsquared:.4f} | Adj. R²: {ols_model.rsquared_adj:.4f} | N: {int(ols_model.nobs)}")

    # HAC inference
    X = sm.add_constant(df[[f"S_lag{k}" for k in range(L+1)]])
    hac = sm.OLS(df["y"], X).fit(cov_type="HAC", cov_kwds={"maxlags":4})
    coefs = hac.params.drop("const"); tvals = hac.tvalues.drop("const"); pvals = hac.pvalues.drop("const")
    table = pd.DataFrame({"coef_%perbp": coefs, "t_hac": tvals, "p_hac": pvals})
    print(f"\n--- Multi-lag HAC coefs: {asset_col} ~ shocked {label} ΔOAS ---")
    print(table.to_string())

    # plot
    plt.figure(figsize=(10,5))
    coefs.plot(kind="bar", title=f"{asset_col} ~ shocked {label} ΔOAS (lags 0–{L})")
    plt.axhline(0, ls="--", c="k"); plt.ylabel("% per 1 bp")
    plt.show()

for lab, Sname in [("HY","HY_OAS_ShockSize"),("IG","IG_OAS_ShockSize")]:
    S = data[Sname]
    for asset in ASSET_LIST:
        fit_multilag(asset, S, lab, L=12)













```
