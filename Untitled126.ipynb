```python
# ============================================================
# UPDATED: adds GOLD ETF @ 30% to EACH existing portfolio
# + fixes TRR (CAGR/AnnualVol), adds correlation tables
# + ensures ALL dates match (inner-join) and uses RETURNS (monthly)
#
# Files expected in /mnt/data:
#   - "STXEMGJ Historical Data.csv"   (EM)
#   - "SYGWDJ Historical Data.csv"    (Other)
#   - "GLDJ Historical Data.csv"      (Gold)
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

EM_FILE    = "STXEMGJ Historical Data.csv"
OTHER_FILE = "SYGWDJ Historical Data.csv"
GOLD_FILE  = "GLDJ Historical Data.csv"

EM_NAME    = "STXEMGJ (EM)"
OTHER_NAME = "SYGWDJ (Other)"
GOLD_NAME  = "GLDJ (Gold)"

OUTLIER_RETURN_THRESHOLD = 0.20  # 20% monthly move flag


# ----------------------------
# Loaders / helpers
# ----------------------------
def _parse_price_series(s: pd.Series) -> pd.Series:
    return (
        s.astype(str)
         .str.replace(",", "", regex=False)
         .str.strip()
         .replace({"": np.nan, "nan": np.nan})
         .astype(float)
    )

def load_investing_csv(path: str, label: str) -> pd.DataFrame:
    df = pd.read_csv(path)
    required = {"Date", "Price"}
    missing = required - set(df.columns)
    if missing:
        raise ValueError(f"{label}: missing required columns {missing}. Found: {list(df.columns)}")

    df["Date"] = pd.to_datetime(df["Date"], format="%Y/%m/%d", errors="coerce")
    df["Price"] = _parse_price_series(df["Price"])
    df = df.dropna(subset=["Date", "Price"]).copy()
    df = df.sort_values("Date").reset_index(drop=True)

    # RETURNS (monthly) from price levels
    df["ret"] = df["Price"].pct_change()
    df["is_outlier"] = df["ret"].abs() > OUTLIER_RETURN_THRESHOLD
    return df[["Date", "Price", "ret", "is_outlier"]]

def equity_curve(r: pd.Series) -> pd.Series:
    return (1.0 + r.fillna(0.0)).cumprod()

def drawdown_series(r: pd.Series) -> pd.Series:
    eq = equity_curve(r)
    peak = eq.cummax()
    return (eq / peak) - 1.0

def max_drawdown_from_returns(r: pd.Series) -> float:
    return float(drawdown_series(r).min())

def metrics_from_monthly_returns(r: pd.Series) -> dict:
    r = r.dropna()
    months = int(len(r))
    if months < 3:
        return {
            "CAGR": np.nan, "VaR_95": np.nan, "Max_Drawdown": np.nan,
            "Annual_Vol": np.nan, "Monthly_Vol": np.nan,
            "TRR_CAGR_over_AnnualVol": np.nan,
            "Months": months
        }

    years = months / 12.0
    eq_end = float((1.0 + r).prod())
    cagr = eq_end ** (1.0 / years) - 1.0

    monthly_vol = float(r.std(ddof=1))
    annual_vol = monthly_vol * np.sqrt(12.0)

    # Historical VaR 95% (monthly) = 5th percentile
    var_95 = float(np.quantile(r, 0.05))

    mdd = max_drawdown_from_returns(r)

    # TRR exactly as you defined (no tricks): CAGR / annual vol
    trr = (cagr / annual_vol) if (np.isfinite(annual_vol) and annual_vol != 0) else np.nan

    return {
        "CAGR": cagr,
        "VaR_95": var_95,
        "Max_Drawdown": mdd,
        "Annual_Vol": annual_vol,
        "Monthly_Vol": monthly_vol,
        "TRR_CAGR_over_AnnualVol": trr,
        "Months": months
    }


# ----------------------------
# Load + align ALL THREE ETFs on identical dates
# ----------------------------
em    = load_investing_csv(EM_FILE, EM_NAME).rename(columns={"Price":"px_em","ret":"r_em","is_outlier":"out_em"})
other = load_investing_csv(OTHER_FILE, OTHER_NAME).rename(columns={"Price":"px_other","ret":"r_other","is_outlier":"out_other"})
gold  = load_investing_csv(GOLD_FILE, GOLD_NAME).rename(columns={"Price":"px_gold","ret":"r_gold","is_outlier":"out_gold"})

df = em.merge(other, on="Date", how="inner").merge(gold, on="Date", how="inner").sort_values("Date").reset_index(drop=True)

print("Common sample window (inner-joined across ALL 3):")
print("Start:", df["Date"].min(), "End:", df["Date"].max(), "Rows:", len(df))

# Flag obvious funky numbers (big monthly moves)
outliers = df[(df["out_em"]) | (df["out_other"]) | (df["out_gold"])][["Date","r_em","r_other","r_gold","px_em","px_other","px_gold"]]
if len(outliers) > 0:
    print(f"\nFLAGGED OUTLIERS (|monthly return| > {OUTLIER_RETURN_THRESHOLD:.0%}):")
    print(outliers.to_string(index=False))
else:
    print("\nNo outliers flagged at the chosen threshold.")

# (Optional extra aggressive flag)
extreme = df[(df["r_em"].abs() > 0.35) | (df["r_other"].abs() > 0.35) | (df["r_gold"].abs() > 0.35)][["Date","r_em","r_other","r_gold"]]
if len(extreme) > 0:
    print("\nEXTREME MOVES (>35% monthly)  inspect these rows:")
    print(extreme.to_string(index=False))


# ----------------------------
# Build portfolios (RETURNS)
# ----------------------------
strategies = {}

# Base (2-asset)
strategies["50/50 (EM/Other)"]            = 0.50 * df["r_em"] + 0.50 * df["r_other"]
strategies["60/40 (EM=40%, Other=60%)"]   = 0.40 * df["r_em"] + 0.60 * df["r_other"]
strategies[f"100% {EM_NAME}"]             = df["r_em"]
strategies[f"100% {OTHER_NAME}"]          = df["r_other"]

# Add Gold @ 30% to EACH existing portfolio: New = 70% Old + 30% Gold
gold_w = 0.30
for k in list(strategies.keys()):
    strategies[f"{k} + 30% Gold"] = (1 - gold_w) * strategies[k] + gold_w * df["r_gold"]

# Also include 100% Gold as baseline (useful)
strategies[f"100% {GOLD_NAME}"] = df["r_gold"]


# ----------------------------
# Metrics table
# ----------------------------
rows = []
for name, r in strategies.items():
    m = metrics_from_monthly_returns(r)
    rows.append({"Strategy": name, **m})

metrics = pd.DataFrame(rows)

print("\n=== PERFORMANCE METRICS (monthly returns; annualized where noted) ===")
display(metrics.style.format({
    "CAGR": "{:.2%}",
    "VaR_95": "{:.2%}",
    "Max_Drawdown": "{:.2%}",
    "Annual_Vol": "{:.2%}",
    "Monthly_Vol": "{:.2%}",
    "TRR_CAGR_over_AnnualVol": "{:.2f}",
    "Months": "{:d}",
}))

# ----------------------------
# Correlation coefficients
# ----------------------------
# 1) ETF return correlations
etf_rets = pd.DataFrame({
    EM_NAME: df["r_em"],
    OTHER_NAME: df["r_other"],
    GOLD_NAME: df["r_gold"],
}).dropna()

print("\n=== CORRELATION (ETFs, monthly returns) ===")
display(etf_rets.corr().style.format("{:.3f}"))

# 2) Portfolio return correlations
port_rets = pd.DataFrame({k: v for k, v in strategies.items()}).dropna()

print("\n=== CORRELATION (PORTFOLIOS, monthly returns) ===")
display(port_rets.corr().style.format("{:.3f}"))


# ----------------------------
# Equity curves (PORTFOLIOS) + drawdowns + hist + rolling vol
# ----------------------------
# Equity curves
plt.figure(figsize=(13, 6))
for name, r in strategies.items():
    plt.plot(df["Date"], equity_curve(r), label=name)
plt.title("Equity Curves (Rebased to 1.0)  Portfolios")
plt.xlabel("Date")
plt.ylabel("Growth of 1.0")
plt.legend(ncol=2)
plt.grid(True, alpha=0.3)
plt.show()

# Drawdowns
plt.figure(figsize=(13, 6))
for name, r in strategies.items():
    plt.plot(df["Date"], drawdown_series(r), label=name)
plt.title("Drawdowns  Portfolios")
plt.xlabel("Date")
plt.ylabel("Drawdown")
plt.legend(ncol=2)
plt.grid(True, alpha=0.3)
plt.show()

# Return histograms
plt.figure(figsize=(13, 6))
for name, r in strategies.items():
    plt.hist(r.dropna().values, bins=25, alpha=0.45, label=name)
plt.title("Monthly Return Distributions  Portfolios")
plt.xlabel("Monthly return")
plt.ylabel("Frequency")
plt.legend(ncol=2)
plt.grid(True, alpha=0.3)
plt.show()

# Rolling 12M annualized vol (from monthly returns)
plt.figure(figsize=(13, 6))
for name, r in strategies.items():
    roll_ann_vol = r.rolling(12).std(ddof=1) * np.sqrt(12.0)
    plt.plot(df["Date"], roll_ann_vol, label=name)
plt.title("Rolling 12M Annualized Volatility (monthly returns)  Portfolios")
plt.xlabel("Date")
plt.ylabel("Annualized vol")
plt.legend(ncol=2)
plt.grid(True, alpha=0.3)
plt.show()

```
