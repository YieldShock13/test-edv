```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

file_path = "SPY Historical Data (8).csv"

df =pd.read_csv(file_path)
df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values("Date")

prices = df["Price"]
returns = np.log(prices/prices.shift(1)).dropna()

mean_return = returns.mean ()
std_return = returns.std ()

z_scores = (returns - mean_return)/std_return

bucket_width = 0.5
bins = np.arange(z_scores.min(), z_scores.max () + bucket_width, bucket_width)

plt.hist(z_scores, bins = bins, density=True, color='black')

x=np.linspace(z_scores.min(),z_scores.max(), 5000)
normal_pdf = (1/np.sqrt(2*np.pi))*np.exp(-0.5*x**2)

plt.plot(x, normal_pdf)
plt.xlabel("sdev")
plt.ylabel("pdf")
plt.title("pdf vs normal distro")
plt.show()

stats.probplot(z_scores, dist="norm", plot=plt)
plt.title("Q-Q Plot of Standard Returns")
plt.show()





```
