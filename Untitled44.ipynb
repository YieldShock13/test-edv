```python
# STEP 1: Load CSVs
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patheffects as path_effects
import matplotlib.dates as mdates

us = pd.read_csv("United States 10-Year Bond Yield Historical Data (4).csv")
uk = pd.read_csv("United Kingdom 10-Year Bond Yield Historical Data (6).csv")
germany = pd.read_csv("Germany 10-Year Bond Yield Historical Data (2).csv")
japan = pd.read_csv("Japan 10-Year Bond Yield Historical Data (2).csv")

def clean(df, colname):
    df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
    df = df.dropna(subset=["Date"])
    return df[["Date", "Price"]].rename(columns={"Price": colname})

us = clean(us, "US")
uk = clean(uk, "UK")
germany = clean(germany, "Germany")
japan = clean(japan, "Japan")

# STEP 2: Merge and filter
df = us.merge(uk, on="Date").merge(germany, on="Date").merge(japan, on="Date")
df = df.sort_values("Date")
for col in ["US", "UK", "Germany", "Japan"]:
    df[col] = pd.to_numeric(df[col], errors="coerce")

start_date = pd.to_datetime("2021-07-27")
end_date = pd.to_datetime("2025-07-27")
df = df[(df["Date"] >= start_date) & (df["Date"] <= end_date)]

# Compute spreads
df["UK - US"] = df["UK"] - df["US"]
df["Germany - US"] = df["Germany"] - df["US"]
df["Japan - US"] = df["Japan"] - df["US"]

# STEP 3: Plotting with TRUE quarterly x-axis labels only
font = "DejaVu Sans"
watermark = "Pearl Gray Equity and Research"
spreads = {
    "UK - US": "UK vs US 10Y Yield Spread",
    "Germany - US": "Germany vs US 10Y Yield Spread",
    "Japan - US": "Japan vs US 10Y Yield Spread"
}

# Create quarterly label locations
quarter_ends = pd.date_range(start=start_date, end=end_date, freq="Q")

for col, title in spreads.items():
    plt.figure(figsize=(10, 6))
    plt.plot(df["Date"], df[col], color="white", linewidth=2)

    # Label final value
    final_val = df[col].iloc[-1]
    final_date = df["Date"].iloc[-1]
    plt.text(final_date, final_val, f"{final_val:.2f}%", fontsize=11, color='white',
             ha='right', va='bottom',
             path_effects=[path_effects.withStroke(linewidth=1.5, foreground='black')])

    # Titles and labels
    plt.title(title, fontsize=14, fontweight='bold', fontname=font, color='white')
    plt.xlabel("Date", fontsize=12, fontname=font, color='white')
    plt.ylabel("Spread (%)", fontsize=12, fontname=font, color='white')
    plt.grid(True, linestyle='--', alpha=0.3)
    plt.tick_params(colors='white')

    #  X-axis shows only true quarter-end labels
    plt.xticks(quarter_ends)
    plt.gca().set_xticklabels(
        [d.strftime('%Y-%b') for d in quarter_ends],
        fontsize=10,
        fontname=font,
        color='white',
        rotation=45
    )

    # Style
    plt.gca().set_facecolor("black")
    plt.gcf().patch.set_facecolor("black")

    # Watermark
    plt.text(0.01, 0.98, watermark, transform=plt.gca().transAxes,
             fontsize=9, color='gray', alpha=0.7, ha='left', va='top', fontname=font)

    plt.tight_layout()
    plt.show()
```
