```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd
from numpy import zeros, maximum, power, sqrt, around

#1) building parameters for binomial
spot = 100
steps = 4
sigma = 0.20
rate = 0.05
time = 1
dt = time / steps
u = 1 + sigma * sqrt(dt)
d = 1 - sigma * sqrt(dt)
p = 0.5 + rate * sqrt(dt) / (2 * sigma)
df = 1 / (1 + rate * dt)

#2) user-defined frameworks
def binomial_option(
    spot: float,
    strike: float,
    rate: float,
    sigma: float,
    time: float,
    steps: int,
    output: int = 0,
) -> np.ndarray:

    #3) initialise arrays
    px = zeros((steps + 1, steps + 1))
    cp = zeros((steps + 1, steps + 1))
    V = zeros((steps + 1, steps + 1))
    delta = zeros((steps + 1, steps + 1))

    #4.1) Binomial Loop
    #4.2) Forward Loop
    for j in range(steps + 1):
        for i in range(j + 1):
            px[i, j] = spot * power(u, j - i) * power(d, i)
            cp[i, j] = maximum(px[i, j] - strike, 0)

    #Reverse Loop
    for j in range(steps, -1, -1):
        for i in range(j + 1):

            if j == steps:
                V[i, j] = cp[i, j]
                delta[i, j] = 1.0 if px[i, j] >= strike else 0.0

            else:
                V[i, j] = df * (
                    p * V[i, j + 1] +
                    (1 - p) * V[i + 1, j + 1]
                )

                delta[i, j] = (
                    V[i, j + 1] - V[i + 1, j + 1]
                ) / (
                    px[i, j + 1] - px[i + 1, j + 1]
                )

    results = (
        around(px, 2),
        around(cp, 2),
        around(delta, 4),
        around(V, 2),
    )

    return results[output]


# Test Call to trigger code execution and display output
px_tree = binomial_option(spot=100, strike=100, rate=0.05, sigma=0.20, time=1, steps=4, output=0)
v_tree = binomial_option(spot=100, strike=100, rate=0.05, sigma=0.20, time=1, steps=4, output=3)

print("Binomial Tree:")

fig, ax = plt.subplots(figsize=(8, 5))
for j in range(steps + 1):
    for i in range(j + 1):
        x = j
        y = (j / 2) - i

        ax.scatter(x, y, color='white', s=800, zorder=3)

        label = f"S:{px_tree[i, j]:.1f}\nV:{v_tree[i, j]:.1f}"
        ax.text(x, y, label, ha='center', va='center', color='black', fontsize=7, fontweight='bold')

        if j < steps:
            ax.plot([x, x + 1], [y, y + 0.5], color='black', alpha=0.3, zorder=1)
            ax.plot([x, x + 1], [y, y - 0.5], color='black', alpha=0.3, zorder=1)

ax.axis('off')
plt.show()













```
