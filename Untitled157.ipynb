```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

RF_ANNUAL = 0.045
USDZAR_SPOT_T0 = 16.04

# =========================
# Helpers
# =========================

def to_float(series):
    return (
        series.astype(str)
        .str.replace(",", "", regex=False)
        .str.replace("%", "", regex=False)
        .str.strip()
        .replace({"": np.nan})
        .astype(float)
    )

def load_price(file):
    df = pd.read_csv(file)
    df.columns = [c.strip() for c in df.columns]
    df["Date"] = pd.to_datetime(df["Date"])
    df = df.sort_values("Date").set_index("Date")
    df["Price"] = to_float(df["Price"])
    df = df.resample("ME").last()
    return df["Price"].pct_change(fill_method=None)

def load_monthly_return(file):
    df = pd.read_csv(file)
    df.columns = [c.strip() for c in df.columns]
    df["Date"] = pd.to_datetime(df["Date"])
    df = df.sort_values("Date").set_index("Date")

    # detect return column
    for col in ["Change %","Return","Monthly Return","Return %","Change"]:
        if col in df.columns:
            r = to_float(df[col])
            if r.abs().median(skipna=True) > 1:
                r = r / 100
            return r.resample("ME").last()
    raise ValueError("Could not find Laurium return column.")

# =========================
# Load Assets
# =========================

TRUW     = load_price("TRUJ Historical Data (2).csv")
NEWGOLD  = load_price("GLDJ Historical Data (2).csv")
MSCI_ZAR = load_price("0P0000ZSS4 Historical Data.csv")
MSCI_EM  = load_price("STXEMGJ Historical Data (1).csv")
ALBI     = load_price("0P00015WR7 Historical Data.csv")
URTH     = load_price("URTH Historical Data.csv")
IEF      = load_price("IEF Historical Data (6).csv")
NFLX     = load_price("NFLX Historical Data.csv")
BTC      = load_price("Bitcoin Historical Data (1).csv")

USDZAR   = load_price("USD_ZAR Historical Data (5).csv")
EURUSD   = load_price("EUR_USD Historical Data (1).csv")

PERE1    = load_price("0P0001K7XM Historical Data (2).csv")
PERE2    = load_price("0P0001K7XN Historical Data.csv")

LAURIUM  = load_monthly_return("Laurium_Market_Neutral_Monthly_Returns_Investing_Format.csv")

# =========================
# Dividend / Coupon Overlay
# =========================

TRUW += 0.085/12
IEF  += 0.04/12

# =========================
# Correct FX Math
# =========================

def usd_to_zar(asset):
    return (1+asset)*(1+USDZAR)-1

def eur_to_usd(asset):
    return (1+asset)*(1+EURUSD)-1

def eur_to_zar(asset):
    return (1+eur_to_usd(asset))*(1+USDZAR)-1

def zar_to_usd(asset):
    return (1+asset)/(1+USDZAR)-1

# =========================
# Position Sizes (ZAR NAV)
# =========================

positions_zar = {
    "TRUW":15779,
    "NEWGOLD":21977,
    "MSCI_ZAR":44490,
    "MSCI_EM":26552,
    "ALBI":29332,
    "PERE1":25000,
    "PERE2":25000,
    "LAURIUM":10000,
    "URTH":24060,
    "IEF_EUR":51328,
    "BTC":19793.36,
    "NFLX":6191,
    "ZAR_CASH":30000,
    "USD_CASH":4812
}

# =========================
# Engine
# =========================

def run_engine(base="ZAR", include_btc=True):

    pos = positions_zar.copy()
    if not include_btc:
        pos["BTC"] = 0

    if base == "USD":
        for k in pos:
            pos[k] /= USDZAR_SPOT_T0

    pos = pd.Series(pos)
    weights = pos / pos.sum()

    if base == "ZAR":
        data = pd.concat([
            TRUW,
            NEWGOLD,
            MSCI_ZAR,
            MSCI_EM,
            ALBI,
            PERE1,
            PERE2,
            LAURIUM,
            usd_to_zar(URTH),
            eur_to_zar(IEF),
            usd_to_zar(BTC),
            usd_to_zar(NFLX)
        ], axis=1)
    else:
        data = pd.concat([
            zar_to_usd(TRUW),
            zar_to_usd(NEWGOLD),
            zar_to_usd(MSCI_ZAR),
            zar_to_usd(MSCI_EM),
            zar_to_usd(ALBI),
            zar_to_usd(PERE1),
            zar_to_usd(PERE2),
            zar_to_usd(LAURIUM),
            URTH,
            eur_to_usd(IEF),
            BTC,
            NFLX
        ], axis=1)

    data.columns = [
        "TRUW","NEWGOLD","MSCI_ZAR","MSCI_EM","ALBI",
        "PERE1","PERE2","LAURIUM","URTH","IEF_EUR","BTC","NFLX"
    ]

    # Cash
    if base=="ZAR":
        data["ZAR_CASH"] = 0.06/12
        data["USD_CASH"] = usd_to_zar(pd.Series(0.03/12,index=data.index))
    else:
        data["USD_CASH"] = 0.03/12
        data["ZAR_CASH"] = zar_to_usd(pd.Series(0.06/12,index=data.index))

    data = data.dropna()
    weights = weights.reindex(data.columns).fillna(0)
    weights = weights / weights.sum()

    port_ret = (data * weights).sum(axis=1)

    # ================= Metrics =================

    cagr = (1+port_ret).prod()**(12/len(port_ret))-1
    vol = port_ret.std()*np.sqrt(12)
    sharpe = (port_ret.mean()*12 - RF_ANNUAL)/vol

    cum = (1+port_ret).cumprod()
    max_dd = (cum/cum.cummax()-1).min()
    calmar = cagr/abs(max_dd)

    var95 = -np.percentile(port_ret,5)
    var99 = -np.percentile(port_ret,1)

    bench = data["URTH"]
    beta = np.cov(port_ret,bench)[0,1]/np.var(bench)
    alpha = (port_ret.mean()-beta*bench.mean())*12
    active = port_ret - bench
    ir = (active.mean()*12)/(active.std()*np.sqrt(12))

    cond_mask = bench < 0
    cond_beta = np.cov(port_ret[cond_mask],bench[cond_mask])[0,1]/np.var(bench[cond_mask])

    # Risk Contribution
    cov = data.cov()*12
    w = weights.values
    port_vol = np.sqrt(w.T @ cov.values @ w)
    marginal = cov.values @ w / port_vol
    component = w * marginal
    rc_raw = pd.Series(component,index=data.columns)
    rc_pct = rc_raw / rc_raw.sum()

    print("\n==============================")
    print(f"BASE: {base} | BTC: {include_btc}")
    print("==============================")
    print("CAGR:",cagr)
    print("Vol:",vol)
    print("Sharpe:",sharpe)
    print("Calmar:",calmar)
    print("MaxDD:",max_dd)
    print("VaR:",{0.95:var95,0.99:var99})
    print("Alpha:",alpha)
    print("Beta:",beta)
    print("IR:",ir)
    print("Conditional Beta:",cond_beta)
    print("\nRisk Contribution %:")
    print(rc_pct.sort_values(ascending=False))
    print("Sum RC%:",rc_pct.sum())

    plt.figure()
    cum.plot(title=f"Equity Curve ({base}, BTC={include_btc})")
    plt.show()

    plt.figure()
    (cum/cum.cummax()-1).plot(title=f"Drawdown ({base}, BTC={include_btc})")
    plt.show()

# ================= Run All =================

run_engine("ZAR",True)
run_engine("ZAR",False)
run_engine("USD",True)
run_engine("USD",False)




```
