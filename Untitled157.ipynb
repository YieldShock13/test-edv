```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# =====================================================
# GLOBAL SETTINGS
# =====================================================

START_DATE = "2020-02-01"
USDZAR_SPOT_T0 = 16.04

# Cash
ZAR_CASH = 30000
USD_CASH = 300
ZAR_RATE_ANNUAL = 0.06
USD_RATE_ANNUAL = 0.03

# Dividends / coupons
TRUW_DIV_ANNUAL = 0.085
IEF_COUPON_ANNUAL = 0.04

# Risk Free
RF_ANNUAL = 0.045

# =====================================================
# CONVERSIONS
# =====================================================

def annual_to_monthly(r):
    return (1+r)**(1/12)-1

ZAR_RATE = annual_to_monthly(ZAR_RATE_ANNUAL)
USD_RATE = annual_to_monthly(USD_RATE_ANNUAL)
TRUW_DIV = TRUW_DIV_ANNUAL/12
IEF_COUPON = IEF_COUPON_ANNUAL/12
RF_MONTHLY = annual_to_monthly(RF_ANNUAL)

# =====================================================
# LOADERS
# =====================================================

def load_price_monthly(filename):
    df = pd.read_csv(filename)
    df["Date"] = pd.to_datetime(df["Date"])
    df = df.sort_values("Date")
    df["Price"] = df["Price"].astype(str).str.replace(",","",regex=False).astype(float)
    df = df.set_index("Date")[["Price"]]
    df = df.resample("M").last()
    df["ret"] = df["Price"].pct_change()
    return df["ret"]

def load_monthly_return_file(filename):
    df = pd.read_csv(filename)
    df["Date"] = pd.to_datetime(df["Date"])
    df = df.sort_values("Date")
    df["Change %"] = df["Change %"].astype(str).str.replace("%","").astype(float)/100
    return df.set_index("Date")["Change %"]

# =====================================================
# LOAD ALL SERIES
# =====================================================

truw     = load_price_monthly("TRUJ Historical Data (2).csv") + TRUW_DIV
newgold  = load_price_monthly("GLDJ Historical Data (2).csv")
msci_zar = load_price_monthly("0P0000ZSS4 Historical Data.csv")
msci_em  = load_price_monthly("STXEMGJ Historical Data (1).csv")
albi     = load_price_monthly("0P00015WR7 Historical Data.csv")

urth     = load_price_monthly("URTH Historical Data.csv")
nflx     = load_price_monthly("NFLX Historical Data.csv")
btc      = load_price_monthly("Bitcoin Historical Data (1).csv")
ief      = load_price_monthly("IEF Historical Data (6).csv") + IEF_COUPON

usdzar   = load_price_monthly("USD_ZAR Historical Data (5).csv")
eurusd   = load_price_monthly("EUR_USD Historical Data (1).csv")

pere1    = load_price_monthly("0P0001K7XM Historical Data (2).csv")
pere2    = load_price_monthly("0P0001K7XN Historical Data.csv")
laurium  = load_monthly_return_file("Laurium_Market_Neutral_Monthly_Returns_Investing_Format.csv")

# =====================================================
# ALIGN
# =====================================================

data = pd.concat(
    [truw,newgold,msci_zar,msci_em,albi,
     urth,nflx,btc,ief,
     usdzar,eurusd,
     pere1,pere2,laurium],
    axis=1
)

data.columns = [
    "TRUW","NEWGOLD","MSCI_ZAR","MSCI_EM","ALBI",
    "URTH","NFLX","BTC","IEF",
    "USDZAR","EURUSD",
    "PERE1","PERE2","LAURIUM"
]

data = data.loc[START_DATE:].dropna()

# =====================================================
# FX HELPERS
# =====================================================

def usd_to_zar(r):
    return (1+r)*(1+data["USDZAR"])-1

def zar_to_usd(r):
    return (1+r)/(1+data["USDZAR"])-1

# =====================================================
# BUILD RETURN MATRIX
# =====================================================

def build_returns(base="ZAR"):

    R = pd.DataFrame(index=data.index)
    ief_eur = (1+data["IEF"])/(1+data["EURUSD"])-1

    if base=="ZAR":
        R["TRUW"]=data["TRUW"]
        R["NEWGOLD"]=data["NEWGOLD"]
        R["MSCI_ZAR"]=data["MSCI_ZAR"]
        R["MSCI_EM"]=data["MSCI_EM"]
        R["ALBI"]=data["ALBI"]
        R["PERE1"]=data["PERE1"]
        R["PERE2"]=data["PERE2"]
        R["LAURIUM"]=data["LAURIUM"]

        R["URTH"]=usd_to_zar(data["URTH"])
        R["NFLX"]=usd_to_zar(data["NFLX"])
        R["BTC"]=usd_to_zar(data["BTC"])

        eurzar=(1+data["EURUSD"])*(1+data["USDZAR"])-1
        R["IEF_EUR"]=(1+ief_eur)*(1+eurzar)-1

        R["ZAR_CASH"]=ZAR_RATE
        R["USD_CASH"]=usd_to_zar(USD_RATE)

    if base=="USD":
        R["TRUW"]=zar_to_usd(data["TRUW"])
        R["NEWGOLD"]=zar_to_usd(data["NEWGOLD"])
        R["MSCI_ZAR"]=zar_to_usd(data["MSCI_ZAR"])
        R["MSCI_EM"]=zar_to_usd(data["MSCI_EM"])
        R["ALBI"]=zar_to_usd(data["ALBI"])
        R["PERE1"]=zar_to_usd(data["PERE1"])
        R["PERE2"]=zar_to_usd(data["PERE2"])
        R["LAURIUM"]=zar_to_usd(data["LAURIUM"]

        )
        R["URTH"]=data["URTH"]
        R["NFLX"]=data["NFLX"]
        R["BTC"]=data["BTC"]

        R["IEF_EUR"]=(1+ief_eur)*(1+data["EURUSD"])-1

        R["USD_CASH"]=USD_RATE
        R["ZAR_CASH"]=zar_to_usd(ZAR_RATE)

    return R

# =====================================================
# FROZEN NAV
# =====================================================

nav_zar = pd.Series({
    "TRUW":15779,"NEWGOLD":21977,"MSCI_ZAR":44490,"MSCI_EM":26552,
    "ALBI":29332,"PERE1":25000,"PERE2":25000,"LAURIUM":10000,
    "URTH":1500*USDZAR_SPOT_T0,
    "IEF_EUR":3200*USDZAR_SPOT_T0,
    "BTC":1234*USDZAR_SPOT_T0,
    "NFLX":6191,
    "ZAR_CASH":ZAR_CASH,
    "USD_CASH":USD_CASH*USDZAR_SPOT_T0
})

# =====================================================
# CORE ENGINE
# =====================================================

def run(base="ZAR", include_btc=True):

    print("\n==============================")
    print("BASE:",base," | BTC:",include_btc)
    print("==============================")

    R = build_returns(base)

    nav = nav_zar.copy()
    if base=="USD":
        nav = nav / USDZAR_SPOT_T0

    if not include_btc:
        nav["BTC"] = 0

    weights = nav.reindex(R.columns,fill_value=0)
    weights = weights/weights.sum()

    print("\nPOSITION SIZES:")
    print(nav)
    print("\nWEIGHTS:")
    print(weights)

    port = (R*weights).sum(axis=1)

    # PERFORMANCE
    CAGR = (1+port).prod()**(12/len(port))-1
    vol = port.std()*np.sqrt(12)

    equity = (1+port).cumprod()
    dd = equity/equity.cummax()-1
    maxdd = dd.min()

    excess = port - RF_MONTHLY
    sharpe = (excess.mean()*12)/(port.std()*np.sqrt(12))
    calmar = CAGR/abs(maxdd)

    # VaR
    VaR = {
        0.95: -np.percentile(port,5),
        0.99: -np.percentile(port,1)
    }

    # Benchmark stats
    bench = R["URTH"]
    beta = np.cov(port,bench)[0,1]/np.var(bench)
    alpha = (port.mean()*12) - beta*(bench.mean()*12)

    active = port-bench
    tracking = active.std()*np.sqrt(12)
    IR = (active.mean()*12)/tracking

    down = bench<0
    cond_beta = np.cov(port[down],bench[down])[0,1]/np.var(bench[down])

    # Risk Contribution
    cov = R.cov()*12
    w = weights.values
    port_vol = np.sqrt(w.T @ cov.values @ w)
    mrc = cov.values @ w / port_vol
    rc = w*mrc
    rc = pd.Series(rc,index=R.columns)

    print("\nCAGR:",CAGR)
    print("Vol:",vol)
    print("Sharpe:",sharpe)
    print("Calmar:",calmar)
    print("MaxDD:",maxdd)
    print("VaR:",VaR)
    print("Alpha:",alpha)
    print("Beta:",beta)
    print("IR:",IR)
    print("Conditional Beta:",cond_beta)

    print("\nRisk Contribution:")
    print(rc.sort_values(ascending=False))

    # Charts
    equity.plot(title="Equity Curve")
    plt.show()

    dd.plot(title="Drawdown")
    plt.show()

    (equity/((1+bench).cumprod())).plot(title="Active Equity vs URTH")
    plt.show()

    roll_sharpe = (port.rolling(12).mean()*12)/(port.rolling(12).std()*np.sqrt(12))
    roll_sharpe.plot(title="Rolling 12M Sharpe")
    plt.show()

    roll_beta = port.rolling(12).cov(bench)/bench.rolling(12).var()
    roll_beta.plot(title="Rolling 12M Beta")
    plt.show()

# =====================================================
# EXECUTE
# =====================================================

run("ZAR",True)
run("ZAR",False)
run("USD",True)
run("USD",False)
```
